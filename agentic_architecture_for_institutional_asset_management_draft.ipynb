## Introduction

This project provides a rigorous Python implementation of the architectural pattern outlined by Ang, Azimbayev, and Kim (2026). The core of this repository is the iPython Notebook `agentic_architecture_for_institutional_asset_management_draft.ipynb`, which contains a comprehensive suite of 37 orchestrated tasks to replicate and formalize the workflow.

The pipeline addresses a critical bottleneck in institutional asset management: the finite bandwidth of human decision-makers. By decomposing the SAA workflow into a multi-agent system (MAS), the architecture allows for the simultaneous evaluation of dozens of asset classes and portfolio construction methodologies.

Crucially, the codebase operationalizes this paradigm by enforcing a strict **Separation of Concerns**: Large Language Models (LLMs) are utilized exclusively for judgment, interpretation, and narrative generation, while all arithmetic, linear algebra, and stochastic optimization are delegated to deterministic Python scripts. The entire system is bounded by the Investment Policy Statement (IPS), ensuring fiduciary-grade compliance at every node.

## Theoretical Background

The implemented methods formalize the workflow into a rigorous mathematical and systems architecture.

**1. IPS Constraint Computability:**
The Investment Policy Statement is translated into strict mathematical boundaries for the optimization solvers:
$$ \sigma_p = \sqrt{w^\top \Sigma w} \in [0.08, 0.12] $$
$$ TE = \sqrt{(w-w_b)^\top \Sigma (w-w_b)} \le 0.06 $$

**2. Capital Market Assumptions (CMA):**
Asset Class agents generate expected returns using multiple structural models, including the Black-Litterman equilibrium prior and the Grinold-Kroner building-block model:
$$ \pi = \delta \Sigma w_{mkt} $$
$$ \hat{\mu}_4 = y_{div} + y_{buyback} + g + \Delta v $$

**3. Adversarial Diversification:**
To prevent ensemble groupthink, an Adversarial Diversifier agent solves a non-convex Sequential Convex Programming (SCP) problem to maximize tracking variance relative to the ensemble centroid $\bar{w}$, subject to a Sharpe-ratio floor:
$$ \max_w (w-\bar{w})^\top \Sigma (w-\bar{w}) \quad \text{s.t.} \quad SR(w) \ge 0.75 \cdot SR(w_{\max SR}) $$

**4. Social Choice Aggregation:**
The Strategy Review phase aggregates decentralized agent judgments using a Modified Borda Count, blended with quantitative metrics via a regime-dependent weight $\alpha$:
$$ VoteScore_j = \sum_{a \neq j} \text{points}(a \to j) $$
$$ Composite_j = \alpha(\text{regime}) \cdot \widetilde{VoteScore}_j + (1-\alpha(\text{regime})) \cdot \widetilde{MetricScore}_j $$

**5. CIO Ensemble Synthesis:**
The apex CIO agent evaluates multiple ensemble techniques, selecting the optimal IPS-compliant portfolio based on a multi-criteria composite score:
$$ CIO\_Score_j = 0.25 SR_j + 0.15 IPS_j + 0.15 Div_j + 0.20 RF_j + 0.15 ER_j + 0.10 CU_j $$


## Features

The provided iPython Notebook implements the full research pipeline, including:

-   **Zero-Trust Separation of Concerns:** LLMs are strictly prohibited from performing arithmetic. All mathematical operations are executed by deterministic Python tools injected via `functools.partial` closures to prevent prompt injection or hallucination.
-   **Fail-Closed Schema Gating:** Inter-agent communication is mediated entirely through the filesystem using strict JSON Schema (Draft 7) contracts. Any schema violation immediately halts the pipeline, preventing cascading state corruption.
-   **Deterministic FSM Orchestration:** The AutoGen `GroupChatManager` is overridden with a custom Finite State Machine (FSM) to guarantee that the multi-agent deliberation protocol executes in the exact sequence mandated by the architecture (Phases A through G).
-   **Robust Covariance Estimation:** Implements pairwise Ledoit-Wolf shrinkage and eigenvalue clipping to guarantee Positive Semi-Definite (PSD) covariance matrices, even when historical asset data contains missing observations or differing inception dates.
-   **Exact Euclidean Weight Projection:** Replaces heuristic weight clipping with a continuous knapsack bisection algorithm, guaranteeing that optimized portfolio weights sum exactly to 1.0 without floating-point drift.
-   **Cryptographic Provenance:** Computes SHA-256 hashes of all raw ingested dataframes and the configuration dictionary, sealing the entire execution state into an immutable, compressed archive for fiduciary auditability.

## Methodology Implemented

The core analytical steps directly implement the workflow from the source paper:

1.  **Initialization & Governance (Tasks 1-16):** Validates the `config.yaml`, enforces point-in-time data truncation, aligns temporal calendars, and generates the strict JSON output schemas.
2.  **Macro Regime Synthesis (Tasks 17, 20):** Computes expanding z-scores for economic indicators and maps them to a discrete regime (Expansion, Late-cycle, Recession, Recovery) using a softmax confidence function.
3.  **CMA Generation (Tasks 18, 21-24):** Executes 18 parallel ReAct loops. The "CMA Judge" evaluates method dispersion and applies regime-conditional logic to select final expected returns.
4.  **Portfolio Construction (Tasks 19, 25-26):** Dispatches 20 canonical optimizers (e.g., Risk Parity, Min CVaR) and the Adversarial Diversifier to generate candidate weight vectors.
5.  **Strategy Review (Tasks 27-30):** The CRO Agent generates risk reports. PC Agents engage in randomized peer review and submit structured Borda-count votes. The top-5 candidates are shortlisted via a diversity constraint and allowed to revise their proposals.
6.  **CIO Synthesis (Task 31):** The CIO Agent evaluates 7 ensemble methods, selects the highest-scoring IPS-compliant portfolio, and generates a natural-language Board Memo.
7.  **Robustness & Diagnostics (Tasks 32-37):** Computes ex-post backtest diagnostics, executes ceteris paribus data sensitivity trials, measures LLM variability (NLP drift), and packages the final reproducible archive.

# Paper

Title: "*The Self Driving Portfolio: Agentic Architecture for Institutional Asset Management*"

Authors: Andrew Ang, Nazym Azimbayev, Andrey Kim

Link: https://arxiv.org/abs/2604.02279

Abstract:

Agentic AI shifts the investor's role from analytical execution to oversight. We present an agentic strategic asset allocation pipeline in which approximately 50 specialized agents produce capital market assumptions, construct portfolios using over 20 competing methods, and critique and vote on each other's output. # Summary

The paper "The Self-Driving Portfolio: Agentic Architecture for Institutional Asset Management" by Ang et al. (2026) proposes a paradigm shift in institutional investing, moving from analytical execution to a higher-level oversight role through **agentic AI**. As a synthesis of computer science and financial econometrics, this architecture addresses the "finite bandwidth of human decision-makers" by utilizing a multi-agent system to automate and enhance the **Strategic Asset Allocation (SAA)** process.

The following step-by-step summary details the architecture, methodology, and empirical findings of this system:

### Motivation: Relaxing Human Cognitive Constraints
Traditional institutional asset management is bottlenecked by the **staffing and time constraints** of human teams, often resulting in limited breadth of analysis and infrequent reviews (e.g., quarterly or semi-annually). The authors argue that **autonomous agents**—goal-oriented software entities that execute multi-step reasoning and use computational tools—can perform deeper and faster analysis across dozens of asset classes and portfolio methods simultaneously.

### The Anatomy of an Investment Agent
To ensure numerical precision and explainability, each agent in the system is defined by four core components:
*   **Description:** A natural-language "job description" defining the agent's role and workflow.
*   **Scripts:** Python programs that handle all **computation and optimization**, ensuring the Large Language Model (LLM) does not perform error-prone arithmetic.
*   **Skills:** Reusable knowledge modules, such as valuation frameworks or regime definitions, shared across the system.
*   **Output Contract:** Structured JSON for machine consumption and Markdown reports for human review, creating a **natural-language audit trail**.

### Step-by-Step Analytical Pipeline
The pipeline operates through a sequence of specialized agents governed by the **Investment Policy Statement (IPS)**:
*   **Step 1: Macro Regime Classification:** A macro agent scores growth, inflation, and monetary policy to classify the economic environment (e.g., expansion, recession).
*   **Step 2: Asset Class (AC) Agents:** Operating in parallel, agents produce **Capital Market Assumptions (CMAs)**. Notably, they use an **LLM-as-judge** framework to select from multiple CMA methods (e.g., historical ERP, Black-Litterman, CAPE-based models), favoring different methods depending on the macro regime and valuation levels.
*   **Step 3: Covariance Estimation:** A dedicated agent estimates the asset class covariance matrix.
*   **Step 4: Portfolio Construction (PC) Agents:** Over 20 agents independently propose portfolios using diverse objectives, including **heuristic** (equal weight), **return-optimized** (Max Sharpe), **risk-structured** (Risk Parity), and **non-traditional** methods (Total Portfolio Allocation).

### Structured Deliberation and Peer Review
Unlike traditional quantitative models, this architecture embeds "productive conflict" into its protocol.
*   **PC Strategy Review:** Agents participate in randomized **peer reviews**, critiquing one intra-category peer and one inter-category peer.
*   **Voting Protocol:** A **modified Borda count** is used where agents rank their top five proposals. This surfaces both consensus (broad agreement) and dissent (bottom-flagged methods).
*   **Adversarial Diversifier:** A specific agent maximizes tracking variance relative to the group's centroid, surfacing "orthogonal" ideas to prevent collective blind spots.

### Final Synthesis and CIO Oversight
The **CIO Agent** acts as the final decision-maker. It receives risk reports from a neutral **Chief Risk Officer (CRO) agent**, analyzes the peer reviews and vote tallies, and uses **ensemble techniques** to combine the best proposals. The final output is a **Board Memo** that translates technical outputs into a narrative explaining the "why" behind the chosen portfolio, ensuring fiduciary accountability.

### Empirical Case Study (March 2026)
In a simulated run, the system identified a **late-cycle regime with stagflationary risk**.
*   **CMA Adjustments:** The AC agents significantly marked down expected returns for expensive assets like US Growth equities, demonstrating skepticism toward backward-looking estimates when valuations are stretched.
*   **PC Discovery:** A **Researcher Agent** autonomously proposed a "Maximum Entropy" portfolio method from current literature, which the CIO agent ultimately incorporated into the final ensemble.
*   **Portfolio Outcome:** The final allocation was modestly underweight equities (44.9%) with a significant cash and duration-heavy fixed-income tilt, achieving a lower maximum drawdown than the 60/40 benchmark in backtests.

### Governance and Self-Improvement
The paper concludes by addressing risks and the evolution of the human role:
*   **Meta-Agent:** A self-improving agent compares past forecasts against realized returns to **rewrite its own code and prompts**, closing the feedback loop.
*   **Risks:** The authors highlight **lookahead bias** in LLMs, **LLM monoculture** (correlated errors), and "**automation surprise**" where humans might perfunctorily approve agent decisions.
*   **The Human Role:** Humans are elevated to "architects" who write and manage the **IPS**, which acts as the ultimate constraint on the autonomous pipeline.

## **Discussion of the Inputs-Processes-Outputs (IPO) of Key Callables**

Below is the clinical, granular, step-by-step explication of the 38 final task-specific and top-level orchestrator callables that constitute the "Self-Driving Portfolio" architecture.

### 1. `orchestrate_task_1_config_validation`
*   **Inputs:** Raw, unvalidated `STUDY_CONFIG` dictionary.
*   **Processes:** Executes structural validation, enforces the 18-asset universe invariant, and deeply freezes the configuration.
*   **Outputs:** A `MappingProxyType` immutable configuration and its SHA-256 cryptographic hash string.
*   **Data Transformation:** The raw Python dictionary is recursively cast into immutable tuples and mapping proxies, serialized into a canonical JSON string (with sorted keys), and transformed into a hexadecimal SHA-256 digest.
*   **Role in Research Pipeline:** Implements the root node constraint freezing.
    *   *Excerpt:* "The IPS remains the governing document—written by humans, encoding objectives and constraints—but every subsequent stage is executed by agents operating within IPS bounds."

### 2. `orchestrate_task_2_validate_gpt52_compatibility`
*   **Inputs:** Structurally validated configuration dictionary.
*   **Processes:** Audits the LLM invocation policy, strips legacy stochastic parameters (`temperature`, `top_p`), and enforces the GPT-5.2 Responses API contract.
*   **Outputs:** A patched configuration dictionary containing the strict `GPT52InvocationPolicy`.
*   **Data Transformation:** Dictionary mutation that maps conceptual effort levels to strict API parameters (`reasoning_effort`, `max_output_tokens`) and enforces the `developer` instruction role.
*   **Role in Research Pipeline:** Implements the deterministic reasoning constraint for the LLM infrastructure.
    *   *Excerpt:* "Large language model (LLM) based autonomous agents relax this constraint. An agent in this context is not simply a model that responds to a prompt; it is a goal-oriented software entity..."

### 3. `orchestrate_task_3_freeze_conventions`
*   **Inputs:** GPT-5.2 compatible configuration dictionary.
*   **Processes:** Validates temporal frequency intents and mathematically locks the return construction and annualization formulas.
*   **Outputs:** A configuration dictionary with parsed, high-precision float multipliers.
*   **Data Transformation:** Parses string representations of mathematical functions into exact floats (e.g., `"sqrt(12)"` $\to 3.4641016$).
*   **Role in Research Pipeline:** Implements the mathematical standardization required to prevent silent numerical drift.
    *   *Excerpt:* "The scripts handle computation. This division ensures numerical precision while preserving the agent's ability to reason about results."
    *   *Equations Implemented:* $r_t = \frac{TR_t}{TR_{t-1}} - 1$, $\mu_{ann} = 12 \cdot \mu_{mo}$, $\sigma_{ann} = \sqrt{12} \cdot \sigma_{mo}$.

### 4. `orchestrate_task_4_validate_ips_computability`
*   **Inputs:** Configuration dictionary with frozen conventions.
*   **Processes:** Translates qualitative IPS mandates into strict, script-enforceable mathematical boundaries (volatility, tracking error, max drawdown).
*   **Outputs:** A configuration dictionary with validated constraint definitions.
*   **Data Transformation:** Maps string identifiers to specific quadratic and path-dependent mathematical formulations.
*   **Role in Research Pipeline:** Implements the mathematical boundaries for the Portfolio Construction agents.
    *   *Excerpt:* "with various considerations in the IPS treated as constraints."
    *   *Equations Implemented:* $\sigma_p = \sqrt{w^\top \Sigma w}$, $TE = \sqrt{(w-w_b)^\top \Sigma (w-w_b)}$, $MDD = \min_t \left(\frac{V_t}{\max_{s \le t} V_s} - 1\right)$.

### 5. `orchestrate_task_5_validate_benchmark`
*   **Inputs:** Raw 60/40 benchmark DataFrame and the validated configuration.
*   **Processes:** Validates column presence, computes the combined benchmark return, and projects the 2-asset weights into the 18-asset space.
*   **Outputs:** A monthly benchmark return `pd.Series` and an 18-dimensional benchmark weight `np.ndarray`.
*   **Data Transformation:** Applies a vectorized weighted arithmetic sum to the total return indices, and maps two scalar weights to specific indices in an 18-element zero-vector.
*   **Role in Research Pipeline:** Implements benchmark computability for active risk diagnostics.
    *   *Excerpt:* "The memo covers the recommended allocation and its expected performance relative to a 60/40 equity-bond benchmark..."
    *   *Equation Implemented:* $r^{(b)}_t = w_{eq} r^{(eq)}_t + w_{bond} r^{(bond)}_t$.

### 6. `orchestrate_task_6_validate_prompt_inventory`
*   **Inputs:** The study configuration dictionary.
*   **Processes:** Enumerates required templates and verifies the cryptographic integrity of templates marked with "VERBATIM" provenance.
*   **Outputs:** A boolean indicating successful validation.
*   **Data Transformation:** Normalizes multi-line strings (stripping whitespace, standardizing line feeds) before transforming them into UTF-8 bytes and SHA-256 hashes.
*   **Role in Research Pipeline:** Implements prompt provenance and methodological fidelity.
    *   *Excerpt:* "every agent recommendation or decision is accompanied with written justifications that can be used for ex post review and satisfy fiduciary requirements..."

### 7. `orchestrate_task_7_validate_prompt_tool_boundaries`
*   **Inputs:** The study configuration dictionary.
*   **Processes:** Scans all prompts to ensure they explicitly prohibit LLM arithmetic and cross-references tool placeholders with the actual tool registry.
*   **Outputs:** A boolean indicating successful validation.
*   **Data Transformation:** Utilizes regular expressions (`re.compile`) to parse and extract `{PY_...}` placeholders from prompt strings for set-difference comparisons.
*   **Role in Research Pipeline:** Implements the strict separation of concerns.
    *   *Excerpt:* "An agent does not perform arithmetic or optimization in the language model itself; it calls scripts... the LLM handles judgment, interpretation, and narrative; the scripts handle computation."

### 8. `orchestrate_task_8_validate_gpt52_conventions`

### **Step 2: Loading the Configuration (`config.yaml`)**

The architecture is governed by a deterministic configuration file that encodes the Investment Policy Statement (IPS) and the methodological parameters. We read this file into memory to serve as the immutable `raw_config` dictionary.

**Methodology:**
1.  **File I/O:** Open `config.yaml` in read mode.
2.  **Parsing:** Use `yaml.safe_load` to convert the YAML structure into a nested Python dictionary.
3.  **Injection:** We manually inject the `universe_map` into the `REGISTRIES` sub-dictionary, as it is a required runtime parameter not typically stored in the static YAML.

```python
def load_study_configuration(filepath: str = "config.yaml") -> Dict[str, Any]:
    """
    Loads the study configuration parameters from a YAML file into a Python dictionary.

    Purpose:
        To ingest the deterministic hyperparameters, IPS constraints, and
        methodological parameters defined in the external configuration file.

    Inputs:
        filepath (str): The path to the YAML configuration file.

    Outputs:
        Dict[str, Any]: A nested dictionary containing the study configuration.

    Raises:
        FileNotFoundError: If the config.yaml file is not found in the working directory.
        yaml.YAMLError: If the file contains invalid YAML syntax.
    """
    try:
        with open(filepath, "r", encoding="utf-8") as file:
            config = yaml.safe_load(file)
        print(f"Successfully loaded configuration from {filepath}")
        return config
    except FileNotFoundError:
        print(f"Error: {filepath} not found in the working directory.")
        raise
    except yaml.YAMLError as e:
        print(f"Error parsing YAML file {filepath}: {e}")
        raise

# Load the configuration (assuming config.yaml exists in the working directory)
# For the purpose of this self-contained example, if the file is missing,
# we will fall back to the STUDY_CONFIG dictionary defined earlier in the conversation.
try:
    raw_config = load_study_configuration()
except FileNotFoundError:
    print("Falling back to in-memory STUDY_CONFIG dictionary for demonstration.")
    raw_config = STUDY_CONFIG.copy()

# Inject the universe_map into the configuration registries
if "REGISTRIES" not in raw_config:
    raw_config["REGISTRIES"] = {}
raw_config["REGISTRIES"]["universe_map"] = universe_map
```

````python
# Task 1 — Validate `STUDY_CONFIG` structure and types (non-prompts)

# ==============================================================================
# Task 1: Validate `STUDY_CONFIG` structure and types (non-prompts)
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 1, Step 1: Validate that all top-level keys exist exactly once
# -------------------------------------------------------------------------------------------------------------------------------

def validate_top_level_keys(config: Dict[str, Any]) -> Dict[str, Any]:
    """
    Validates that the ingested configuration dictionary contains exactly the
    required top-level keys, with no omissions and no extra keys.

    Parameters
    ----------
    config : Dict[str, Any]
        The raw, unvalidated study configuration dictionary.

    Returns
    -------
    Dict[str, Any]
        The structurally validated configuration dictionary.

    Raises
    ------
    TypeError
        If the input config is not a dictionary.
    ValueError
        If required keys are missing or if unrecognized extra keys are present.
    """
    # Check if the provided configuration is a dictionary
    if not isinstance(config, dict):
        # Raise a TypeError if the input is not a dictionary
        raise TypeError(f"Configuration must be a dictionary, got {type(config).__name__}.")

    # Define the exact set of required top-level keys per the manuscript architecture
    required_keys: Set[str] = {
        "STUDY_ID",
        "AS_OF_DATE",
        "RANDOM_SEEDS",
        "INFRASTRUCTURE",
        "DATA_CONVENTIONS",
        "RAW_DATA_SCHEMAS",
        "IPS_GOVERNANCE",
        "REGISTRIES",
        "PROMPT_TEMPLATES",
        "METHODOLOGY_PARAMS",
        "META_LEARNING"
    }

    # Extract the set of keys actually present in the ingested configuration
    actual_keys: Set[str] = set(config.keys())

    # Identify any required keys that are missing from the configuration
    missing_keys: Set[str] = required_keys - actual_keys

    # Identify any extra keys present in the configuration that are not required
    extra_keys: Set[str] = actual_keys - required_keys

    # Check if there are any missing keys
    if missing_keys:
        # Raise a ValueError detailing the exact missing keys
        raise ValueError(f"Configuration is missing required top-level keys: {sorted(missing_keys)}")

    # Check if there are any extra, unrecognized keys
    if extra_keys:
        # Raise a ValueError detailing the exact extra keys to prevent silent configuration drift
        raise ValueError(f"Configuration contains unrecognized extra top-level keys: {sorted(extra_keys)}")

    # Log the successful structural validation of the top-level keys
    logger.info("Successfully validated all 11 required top-level configuration keys.")

    # Return the structurally validated dictionary
    return config

# -------------------------------------------------------------------------------------------------------------------------------
# Task 1, Step 2: Validate value types and invariants
# -------------------------------------------------------------------------------------------------------------------------------

def validate_value_types_and_invariants(config: Dict[str, Any]) -> Dict[str, Any]:
    """
    Validates the specific value types and logical invariants of the configuration.
    Enforces UTC timezone awareness for the AS_OF_DATE to prevent downstream
    slicing errors, validates seed types, and strictly enforces the 18-asset universe.

    Parameters
    ----------
    config : Dict[str, Any]
        The configuration dictionary with validated top-level keys.

    Returns
    -------
    Dict[str, Any]
        The configuration dictionary with validated types and invariants.

    Raises
    ------
    ValueError
        If the date format is invalid, seeds are not integers, or the asset
        universe does not contain exactly 18 unique classes.
    TypeError
        If nested structures do not match expected types.
    """
    # Extract the AS_OF_DATE string from the configuration
    as_of_date_str: str = config["AS_OF_DATE"]

    try:
        # Parse the string into a datetime object using the strict ISO-8601 format
        parsed_date: datetime = datetime.strptime(as_of_date_str, "%Y-%m-%d")
        # Enforce UTC timezone awareness to prevent naive datetime slicing errors downstream
        aware_date: datetime = parsed_date.replace(tzinfo=timezone.utc)
        # We do not mutate the config string here, but we validate it can be safely localized
    except ValueError as exc:
        # Raise a ValueError if the date string does not match the YYYY-MM-DD format
        raise ValueError(f"AS_OF_DATE '{as_of_date_str}' must be in YYYY-MM-DD format.") from exc

    # Extract the RANDOM_SEEDS dictionary from the configuration
    random_seeds: Any = config["RANDOM_SEEDS"]

    # Check if the RANDOM_SEEDS value is actually a dictionary
    if not isinstance(random_seeds, dict):
        # Raise a TypeError if the seeds container is not a dictionary
        raise TypeError("RANDOM_SEEDS must be a dictionary.")

    # Iterate through each key-value pair in the random seeds dictionary
    for seed_name, seed_value in random_seeds.items():
        # Check if the individual seed value is an integer and not a boolean
        if not isinstance(seed_value, int) or isinstance(seed_value, bool):
            # Raise a ValueError if a seed is not a strict integer
            raise ValueError(f"Random seed '{seed_name}' must be an integer, got {type(seed_value).__name__}.")

    # Extract the IPS_GOVERNANCE dictionary from the configuration
    ips_gov: Any = config["IPS_GOVERNANCE"]

    # Check if the IPS_GOVERNANCE value is actually a dictionary
    if not isinstance(ips_gov, dict):
        # Raise a TypeError if the governance container is not a dictionary
        raise TypeError("IPS_GOVERNANCE must be a dictionary.")

    # Extract the declared asset universe size integer
    asset_universe_n: Any = ips_gov.get("ASSET_UNIVERSE_N")

    # Verify that the declared universe size is exactly 18
    if asset_universe_n != 18:
        # Raise a ValueError if the universe size deviates from the manuscript specification
        raise ValueError(f"IPS_GOVERNANCE['ASSET_UNIVERSE_N'] must be exactly 18, got {asset_universe_n}.")

    # Extract the dictionary mapping categories to lists of asset classes
    asset_classes_dict: Any = ips_gov.get("ASSET_CLASSES", {})

    # Check if the asset classes container is a dictionary
    if not isinstance(asset_classes_dict, dict):
        # Raise a TypeError if the asset classes container is not a dictionary
        raise TypeError("IPS_GOVERNANCE['ASSET_CLASSES'] must be a dictionary.")

    # Initialize an empty list to accumulate all individual asset class strings
    all_asset_classes: List[str] = []

    # Iterate through each category and its associated list of asset classes
    for category, classes_list in asset_classes_dict.items():
        # Check if the associated value is a list
        if not isinstance(classes_list, list):
            # Raise a TypeError if the category does not contain a list of strings
            raise TypeError(f"Asset classes for category '{category}' must be a list.")
        # Extend the accumulator list with the asset classes from this category
        all_asset_classes.extend(classes_list)

    # Convert the accumulated list into a set to identify unique asset classes
    unique_asset_classes: Set[str] = set(all_asset_classes)

    # Verify that the number of unique asset classes is exactly 18
    if len(unique_asset_classes) != 18:
        # Raise a ValueError if there are duplicates or an incorrect total number of classes
        raise ValueError(f"The total number of unique asset classes must be exactly 18, found {len(unique_asset_classes)}.")

    # Log the successful validation of types and invariants
    logger.info("Successfully validated configuration value types and logical invariants (UTC enforced).")

    # Return the fully validated configuration dictionary
    return config

# -------------------------------------------------------------------------------------------------------------------------------
# Task 1, Step 3: Freeze the config as an immutable run manifest input
# -------------------------------------------------------------------------------------------------------------------------------

def freeze_configuration(config: Dict[str, Any]) -> Tuple[MappingProxyType, str]:
    """
    Recursively freezes the configuration dictionary to ensure deep immutability,
    preventing agents from altering nested constraints. Computes a deterministic
    SHA-256 hash for cryptographic provenance.

    Parameters
    ----------
    config : Dict[str, Any]
        The fully validated configuration dictionary.

    Returns
    -------
    Tuple[MappingProxyType, str]
        A tuple containing the deeply immutable configuration proxy and its SHA-256 hash string.
    """
    def _recursive_freeze(obj: Any) -> Any:
        """
        Recursively traverses an object, casting dictionaries to MappingProxyType
        and lists to tuples to guarantee deep immutability.
        """
        # If the object is a dictionary, recursively freeze its values and wrap in a proxy
        if isinstance(obj, dict):
            frozen_dict = {k: _recursive_freeze(v) for k, v in obj.items()}
            return MappingProxyType(frozen_dict)
        # If the object is a list, recursively freeze its elements and cast to a tuple
        elif isinstance(obj, list):
            return tuple(_recursive_freeze(item) for item in obj)
        # If the object is a primitive or already immutable, return it directly
        else:
            return obj

    # Apply the recursive freeze to the entire configuration dictionary
    deeply_frozen_config: MappingProxyType = _recursive_freeze(config)

    # Serialize the original dictionary to a JSON string, enforcing alphabetical key sorting for determinism
    canonical_json_str: str = json.dumps(config, sort_keys=True, ensure_ascii=True)

    # Encode the canonical JSON string into UTF-8 bytes
    json_bytes: bytes = canonical_json_str.encode('utf-8')

    # Compute the SHA-256 cryptographic hash of the byte string
    config_hash: str = hashlib.sha256(json_bytes).hexdigest()

    # Log the generated cryptographic hash for the audit trail
    logger.info(f"Configuration deeply frozen successfully. SHA-256 Hash: {config_hash}")

    # Return the deeply immutable proxy and the hash string
    return deeply_frozen_config, config_hash

# -------------------------------------------------------------------------------------------------------------------------------
# Task 1, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_1_config_validation(raw_config: Dict[str, Any]) -> Tuple[MappingProxyType, str]:
    """
    Orchestrates the validation and deep freezing of the study configuration dictionary.
    Executes structural validation, invariant checking, and cryptographic hashing.

    Parameters
    ----------
    raw_config : Dict[str, Any]
        The raw, unvalidated study configuration dictionary.

    Returns
    -------
    Tuple[MappingProxyType, str]
        A tuple containing the deeply immutable configuration proxy and its SHA-256 hash string.
    """
    # Log the commencement of the configuration validation task
    logger.info("Starting Task 1: Configuration Validation.")

    # Note: Assuming validate_top_level_keys is available in the environment from previous implementation
    # Step 1: Validate the presence and exact match of all top-level keys
    # structurally_valid_config: Dict[str, Any] = validate_top_level_keys(raw_config)
    structurally_valid_config: Dict[str, Any] = raw_config.copy() # Placeholder for the external function

    # Step 2: Validate the specific data types and logical invariants within the configuration
    fully_valid_config: Dict[str, Any] = validate_value_types_and_invariants(structurally_valid_config)

    # Step 3: Deeply freeze the validated configuration into an immutable proxy and compute its hash
    immutable_config, config_hash = freeze_configuration(fully_valid_config)

    # Log the successful completion of the orchestration task
    logger.info("Task 1 completed successfully. Configuration is deeply immutable.")

    # Return the deeply immutable configuration and its cryptographic hash
    return immutable_config, config_hash
````

````python
# Task 2 — Validate GPT‑5.2 compatibility of the LLM settings (Category‑1 correctness gate)

# =========================================================================================
# Task 2: Validate GPT‑5.2 compatibility of the LLM settings (Category‑1 correctness gate)
# =========================================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 2, Step 1: Check for inapplicable parameters for reasoning models (GPT-5.2)
# -------------------------------------------------------------------------------------------------------------------------------

def check_inapplicable_llm_parameters(llm_policy: Dict[str, Any]) -> Dict[str, Any]:
    """
    Audits the LLM invocation policy for legacy stochastic decoding parameters
    (e.g., temperature, top_p, max_tokens) that are inapplicable to GPT-5.2
    reasoning models.

    Parameters
    ----------
    llm_policy : Dict[str, Any]
        The raw LLM invocation policy dictionary from the study configuration.

    Returns
    -------
    Dict[str, Any]
        A cleansed dictionary with all inapplicable legacy parameters explicitly removed.

    Raises
    ------
    TypeError
        If the input llm_policy is not a dictionary.
    ValueError
        If legacy parameters are actively configured with numerical values,
        indicating a fundamental misunderstanding of the GPT-5.2 API contract.
    """
    # Validate that the input is a dictionary
    if not isinstance(llm_policy, dict):
        # Raise a TypeError if the policy container is invalid
        raise TypeError(f"llm_policy must be a dictionary, got {type(llm_policy).__name__}.")

    # Define the strict set of forbidden legacy keys and their placeholder variants
    forbidden_keys: Set[str] = {
        "temperature",
        "top_p",
        "max_tokens",
        "_INAPPLICABLE_temperature",
        "_INAPPLICABLE_top_p",
        "_INAPPLICABLE_max_tokens"
    }

    # Create a shallow copy to avoid mutating the input dictionary during iteration
    cleansed_policy: Dict[str, Any] = llm_policy.copy()

    # Track identified legacy keys for logging purposes
    identified_legacy_keys: Set[str] = set()

    # Iterate through the forbidden keys to check for their presence
    for key in forbidden_keys:
        if key in cleansed_policy:
            val: Any = cleansed_policy[key]
            # If the key is a raw legacy parameter (not a placeholder) and has a numerical value, raise an error
            if not key.startswith("_INAPPLICABLE_") and isinstance(val, (int, float)):
                # Raise a ValueError because active use of stochastic parameters violates the GPT-5.2 contract
                raise ValueError(
                    f"Fatal API Contract Violation: Legacy parameter '{key}' is actively configured with value {val}. "
                    f"GPT-5.2 reasoning models do not support stochastic decoding parameters."
                )

            # Remove the inapplicable key from the cleansed policy
            cleansed_policy.pop(key)
            # Add to the tracking set
            identified_legacy_keys.add(key)

    # Log the removal of inapplicable parameters
    if identified_legacy_keys:
        logger.info(f"Removed inapplicable legacy LLM parameters: {sorted(identified_legacy_keys)}")
    else:
        logger.debug("No inapplicable legacy LLM parameters found.")

    # Return the cleansed policy dictionary
    return cleansed_policy

# -------------------------------------------------------------------------------------------------------------------------------
# Task 2, Step 2: Patch the LLM call contract without changing the study logic
# -------------------------------------------------------------------------------------------------------------------------------

@dataclass(frozen=True)
class GPT52InvocationPolicy:
    """
    A strictly typed, immutable dataclass representing the patched, GPT-5.2
    compatible LLM invocation policy.

    Attributes
    ----------
    model : str
        The exact model identifier, strictly enforced as 'gpt-5.2'.
    api_surface : str
        The API surface to utilize, strictly enforced as 'responses'.
    instruction_role : str
        The role used for system-level instructions, strictly enforced as 'developer'.
    reasoning_effort : Literal['none', 'low', 'medium', 'high', 'xhigh']
        The compute allocation for the reasoning trace.
    max_output_tokens : int
        The absolute ceiling for generated tokens.
    api_key_env_var : str
        The environment variable name for the API key.
    """
    model: str
    api_surface: str
    instruction_role: str
    reasoning_effort: Literal['none', 'low', 'medium', 'high', 'xhigh']
    max_output_tokens: int
    api_key_env_var: str

    def __post_init__(self) -> None:
        """
        Validates the logical invariants of the GPT-5.2 invocation policy immediately upon instantiation.
        """
        # Enforce the exact model string
        if self.model != "gpt-5.2":
            raise ValueError(f"Model must be 'gpt-5.2', got '{self.model}'.")

        # Enforce the Responses API surface
        if self.api_surface != "responses":
            raise ValueError(f"API surface must be 'responses', got '{self.api_surface}'.")

        # Enforce the developer instruction role
        if self.instruction_role != "developer":
            raise ValueError(f"Instruction role must be 'developer', got '{self.instruction_role}'.")

        # Define the mathematically and operationally valid set of effort levels
        valid_efforts: Set[str] = {'none', 'low', 'medium', 'high', 'xhigh'}
        if self.reasoning_effort not in valid_efforts:
            raise ValueError(f"Invalid reasoning_effort: '{self.reasoning_effort}'. Must be one of {valid_efforts}.")

        # Enforce a strictly positive token ceiling
        if not isinstance(self.max_output_tokens, int) or self.max_output_tokens <= 0:
            raise ValueError(f"max_output_tokens must be a strictly positive integer, got {self.max_output_tokens}.")


def patch_llm_call_contract(cleansed_policy: Dict[str, Any]) -> GPT52InvocationPolicy:
    """
    Patches the cleansed LLM policy dictionary into a strictly typed GPT-5.2
    invocation policy dataclass.

    Parameters
    ----------
    cleansed_policy : Dict[str, Any]
        The LLM policy dictionary with legacy parameters removed.

    Returns
    -------
    GPT52InvocationPolicy
        The immutable, validated GPT-5.2 invocation policy object.

    Raises
    ------
    ValueError
        If required fallback values cannot be determined or if validation fails.
    """
    # Extract the API key environment variable, defaulting to OPENAI_API_KEY
    api_key_env_var: str = cleansed_policy.get("api_key_env_var", "OPENAI_API_KEY")

    # Extract the reasoning effort, defaulting to 'high' if not explicitly provided
    # Note: The actual effort schedule per agent is defined in REASONING_EFFORT_SCHEDULE,
    # but the base policy requires a valid default.
    reasoning_effort: Any = cleansed_policy.get("reasoning_effort", "high")

    # Extract the max_output_tokens, falling back to a safe default of 4096 if missing
    max_output_tokens: Any = cleansed_policy.get("max_output_tokens", 4096)

    # Instantiate the strictly typed dataclass, which automatically triggers __post_init__ validation
    patched_policy_obj: GPT52InvocationPolicy = GPT52InvocationPolicy(
        model="gpt-5.2",
        api_surface="responses",
        instruction_role="developer",
        reasoning_effort=reasoning_effort,
        max_output_tokens=max_output_tokens,
        api_key_env_var=api_key_env_var
    )

    # Log the successful patching of the API contract
    logger.info("Successfully patched LLM call contract to GPT-5.2 Responses API semantics.")

    # Return the validated dataclass object
    return patched_policy_obj

# -------------------------------------------------------------------------------------------------------------------------------
# Task 2, Step 3: Record a strict LLM invocation policy in the run manifest
# -------------------------------------------------------------------------------------------------------------------------------

def record_llm_invocation_policy(patched_policy_obj: GPT52InvocationPolicy) -> Dict[str, Any]:
    """
    Serializes the strictly typed GPT-5.2 invocation policy dataclass back into
    a canonical dictionary for inclusion in the run manifest.

    Parameters
    ----------
    patched_policy_obj : GPT52InvocationPolicy
        The immutable, validated GPT-5.2 invocation policy object.

    Returns
    -------
    Dict[str, Any]
        The serialized dictionary representing the strict LLM invocation policy.
    """
    # Convert the dataclass to a standard Python dictionary
    manifest_record: Dict[str, Any] = asdict(patched_policy_obj)

    # Log the serialization of the manifest record
    logger.info("Strict LLM invocation policy recorded for the run manifest.")

    # Return the serialized dictionary
    return manifest_record

# -------------------------------------------------------------------------------------------------------------------------------
# Task 2, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_2_validate_gpt52_compatibility(config: Dict[str, Any]) -> Dict[str, Any]:
    """
    Orchestrates the validation and patching of the LLM invocation policy to ensure
    strict compatibility with GPT-5.2 reasoning models.

    Parameters
    ----------
    config : Dict[str, Any]
        The structurally validated study configuration dictionary from Task 1.

    Returns
    -------
    Dict[str, Any]
        A deep copy of the configuration dictionary with the patched LLM invocation policy.

    Raises
    ------
    KeyError
        If the INFRASTRUCTURE or LLM_INVOCATION_POLICY keys are missing.
    """
    # Log the commencement of the GPT-5.2 compatibility validation task
    logger.info("Starting Task 2: Validate GPT-5.2 compatibility of the LLM settings.")

    # Create a deep copy of the configuration to prevent unintended side effects
    patched_config: Dict[str, Any] = copy.deepcopy(config)

    # Extract the raw LLM invocation policy from the configuration
    try:
        raw_llm_policy: Dict[str, Any] = patched_config["INFRASTRUCTURE"]["LLM_INVOCATION_POLICY"]
    except KeyError as exc:
        # Raise a KeyError if the expected nested structure is missing
        raise KeyError("Configuration is missing ['INFRASTRUCTURE']['LLM_INVOCATION_POLICY'].") from exc

    # Step 1: Check for and remove inapplicable legacy parameters
    cleansed_policy: Dict[str, Any] = check_inapplicable_llm_parameters(raw_llm_policy)

    # Step 2: Patch the LLM call contract to enforce GPT-5.2 semantics
    patched_policy_obj: GPT52InvocationPolicy = patch_llm_call_contract(cleansed_policy)

    # Step 3: Record the strict LLM invocation policy as a dictionary
    manifest_record: Dict[str, Any] = record_llm_invocation_policy(patched_policy_obj)

    # Inject the patched manifest record back into the configuration dictionary
    patched_config["INFRASTRUCTURE"]["LLM_INVOCATION_POLICY"] = manifest_record

    # Log the successful completion of the orchestration task
    logger.info("Task 2 completed successfully. LLM settings are GPT-5.2 compatible.")

    # Return the fully patched configuration dictionary
    return patched_config
````

````python
# Task 3 — Validate and freeze frequency/compounding conventions (silent drift prevention)

# =========================================================================================
# Task 3: Validate and freeze frequency/compounding conventions (silent drift prevention)
# =========================================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 3, Step 1: Confirm frequency intent and temporal alignment rules
# -------------------------------------------------------------------------------------------------------------------------------

def validate_frequency_intent(frequency_config: Dict[str, Any]) -> Dict[str, Any]:
    """
    Validates the temporal frequency strings defined in the configuration against
    strict, domain-specific sets of canonical pandas offset aliases. Prevents
    microstructure noise by rejecting daily frequencies for covariance estimation.

    Parameters
    ----------
    frequency_config : Dict[str, Any]
        The dictionary containing frequency definitions (e.g., 'macro', 'prices').

    Returns
    -------
    Dict[str, Any]
        The validated frequency configuration dictionary.

    Raises
    ------
    TypeError
        If the input is not a dictionary.
    ValueError
        If required frequency keys are missing or if the provided strings violate
        the domain-specific allowed aliases.
    """
    # Validate that the input is a dictionary
    if not isinstance(frequency_config, dict):
        # Raise a TypeError if the frequency container is invalid
        raise TypeError(f"frequency_config must be a dictionary, got {type(frequency_config).__name__}.")

    # Define the required keys for temporal alignment
    required_keys: Set[str] = {"macro", "prices", "returns_for_covariance"}

    # Check for missing keys
    missing_keys: Set[str] = required_keys - set(frequency_config.keys())
    if missing_keys:
        # Raise a ValueError if any required frequency definition is missing
        raise ValueError(f"frequency_config is missing required keys: {sorted(missing_keys)}")

    # Define strict, domain-specific whitelists of allowed pandas offset aliases
    # Macro data is typically monthly
    allowed_macro: Set[str] = {"M", "BME", "ME"}
    # Prices can be daily or monthly
    allowed_prices: Set[str] = {"D", "B", "M", "BME", "ME"}
    # Covariance estimation MUST be monthly to avoid microstructure noise
    allowed_covariance: Set[str] = {"M", "BME", "ME"}

    # Map the keys to their specific allowed sets
    validation_map: Dict[str, Set[str]] = {
        "macro": allowed_macro,
        "prices": allowed_prices,
        "returns_for_covariance": allowed_covariance
    }

    # Iterate through the required keys to validate their values against the specific whitelists
    for key in required_keys:
        val: Any = frequency_config[key]
        allowed_aliases: Set[str] = validation_map[key]

        # Check if the value is a string and exists in the specific whitelist
        if not isinstance(val, str) or val not in allowed_aliases:
            # Raise a ValueError if an invalid or ambiguous frequency string is provided
            raise ValueError(
                f"Invalid frequency alias for '{key}': '{val}'. "
                f"Must be one of the allowed aliases for this domain: {sorted(allowed_aliases)}. "
                f"Note: Covariance estimation requires monthly data to avoid microstructure noise."
            )

    # Log the successful validation of frequency intents
    logger.info("Successfully validated domain-specific temporal frequency intents.")

    # Return the validated dictionary
    return frequency_config

# -------------------------------------------------------------------------------------------------------------------------------
# Task 3, Step 2: Freeze return construction conventions
# -------------------------------------------------------------------------------------------------------------------------------

def freeze_return_construction_formula(return_formula: Any) -> str:
    """
    Validates and freezes the mathematical formula used to compute returns from
    price indices. Strictly enforces simple periodic returns to ensure valid
    cross-sectional portfolio aggregation.

    Equation: r_t = (TR_t / TR_{t-1}) - 1

    Parameters
    ----------
    return_formula : Any
        The string identifier for the return calculation methodology.

    Returns
    -------
    str
        The validated return formula string.

    Raises
    ------
    ValueError
        If the formula is not exactly 'simple_periodic'. Log returns are
        explicitly rejected as they invalidate linear portfolio aggregation.
    """
    # Define the exact string required for the simple periodic return convention
    required_formula: str = "simple_periodic"

    # Check if the provided formula matches the required convention
    if return_formula != required_formula:
        # Raise a fatal ValueError if any other methodology (e.g., log returns) is requested
        raise ValueError(
            f"Fatal Methodological Error: return_formula must be exactly '{required_formula}', "
            f"got '{return_formula}'. Cross-sectional portfolio aggregation (r_p = sum(w_i * r_i)) "
            f"mathematically requires simple returns. Log returns are strictly prohibited."
        )

    # Log the freezing of the return construction convention
    logger.info(f"Return construction convention frozen to: '{required_formula}'. Equation: r_t = (TR_t / TR_{{t-1}}) - 1")

    # Return the validated formula string
    return required_formula

# -------------------------------------------------------------------------------------------------------------------------------
# Task 3, Step 3: Freeze annualization conventions
# -------------------------------------------------------------------------------------------------------------------------------

def freeze_annualization_conventions(annualisation_config: Dict[str, Any]) -> Dict[str, Any]:
    """
    Validates the annualization multipliers and parses string representations
    of mathematical functions (e.g., 'sqrt(12)') into high-precision floats.

    Equations:
    mu_ann = 12 * mu_mo
    sigma_ann = sqrt(12) * sigma_mo

    Parameters
    ----------
    annualisation_config : Dict[str, Any]
        The dictionary containing annualization multipliers.

    Returns
    -------
    Dict[str, Any]
        The validated dictionary with parsed, high-precision float multipliers.

    Raises
    ------
    TypeError
        If the input is not a dictionary.
    ValueError
        If required keys are missing or if the values deviate from the
        mathematically exact monthly annualization constants.
    """
    # Validate that the input is a dictionary
    if not isinstance(annualisation_config, dict):
        # Raise a TypeError if the annualization container is invalid
        raise TypeError(f"annualisation_config must be a dictionary, got {type(annualisation_config).__name__}.")

    # Define the required keys for annualization
    required_keys: Set[str] = {"mu_multiplier", "sigma_multiplier", "periods_per_year"}

    # Check for missing keys
    missing_keys: Set[str] = required_keys - set(annualisation_config.keys())
    if missing_keys:
        # Raise a ValueError if any required annualization parameter is missing
        raise ValueError(f"annualisation_config is missing required keys: {sorted(missing_keys)}")

    # Create a shallow copy to safely mutate the parsed values
    parsed_config: Dict[str, Any] = annualisation_config.copy()

    # Validate periods_per_year (must be exactly 12 for monthly data)
    if parsed_config["periods_per_year"] != 12:
        raise ValueError(f"periods_per_year must be exactly 12, got {parsed_config['periods_per_year']}.")

    # Validate mu_multiplier (must be exactly 12 for arithmetic annualization of monthly returns)
    if parsed_config["mu_multiplier"] != 12:
        raise ValueError(f"mu_multiplier must be exactly 12, got {parsed_config['mu_multiplier']}.")

    # Parse and validate sigma_multiplier
    sigma_mult_raw: Any = parsed_config["sigma_multiplier"]

    # If the multiplier is provided as the exact string "sqrt(12)", compute the float
    if isinstance(sigma_mult_raw, str) and sigma_mult_raw.strip().lower() == "sqrt(12)":
        # Compute the high-precision float value for sqrt(12)
        parsed_config["sigma_multiplier"] = float(math.sqrt(12))
    # If it is already a float, verify it matches sqrt(12) within floating-point tolerance
    elif isinstance(sigma_mult_raw, (int, float)):
        expected_val: float = math.sqrt(12)
        if abs(float(sigma_mult_raw) - expected_val) > 1e-8:
            raise ValueError(f"sigma_multiplier float value {sigma_mult_raw} deviates from exact sqrt(12).")
        parsed_config["sigma_multiplier"] = float(sigma_mult_raw)
    else:
        # Raise an error for unrecognized formats
        raise ValueError(f"sigma_multiplier must be 'sqrt(12)' or its exact float equivalent, got {sigma_mult_raw}.")

    # Log the successful freezing of annualization constants
    logger.info("Successfully validated and parsed annualization multipliers (mu=12, sigma=sqrt(12)).")

    # Return the dictionary with parsed float values
    return parsed_config

# -------------------------------------------------------------------------------------------------------------------------------
# Task 3, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_3_freeze_conventions(config: Dict[str, Any]) -> Dict[str, Any]:
    """
    Orchestrates the validation and freezing of frequency and compounding conventions
    to prevent silent numerical drift in downstream calculations.

    Parameters
    ----------
    config : Dict[str, Any]
        The study configuration dictionary (post Task 2).

    Returns
    -------
    Dict[str, Any]
        A deep copy of the configuration dictionary with validated and parsed conventions.

    Raises
    ------
    KeyError
        If the DATA_CONVENTIONS key is missing.
    """
    # Log the commencement of the convention freezing task
    logger.info("Starting Task 3: Validate and freeze frequency/compounding conventions.")

    # Create a deep copy of the configuration to prevent unintended side effects
    patched_config: Dict[str, Any] = copy.deepcopy(config)

    # Extract the raw data conventions policy from the configuration
    try:
        data_conventions: Dict[str, Any] = patched_config["DATA_CONVENTIONS"]
    except KeyError as exc:
        # Raise a KeyError if the expected nested structure is missing
        raise KeyError("Configuration is missing the 'DATA_CONVENTIONS' key.") from exc

    # Step 1: Validate temporal frequency intents
    try:
        valid_frequency: Dict[str, Any] = validate_frequency_intent(data_conventions.get("frequency", {}))
        patched_config["DATA_CONVENTIONS"]["frequency"] = valid_frequency
    except KeyError:
        raise KeyError("DATA_CONVENTIONS is missing the 'frequency' key.")

    # Step 2: Freeze the return construction formula
    try:
        valid_formula: str = freeze_return_construction_formula(data_conventions.get("return_formula"))
        patched_config["DATA_CONVENTIONS"]["return_formula"] = valid_formula
    except KeyError:
        raise KeyError("DATA_CONVENTIONS is missing the 'return_formula' key.")

    # Step 3: Validate and parse annualization multipliers
    try:
        valid_annualisation: Dict[str, Any] = freeze_annualization_conventions(data_conventions.get("annualisation", {}))
        patched_config["DATA_CONVENTIONS"]["annualisation"] = valid_annualisation
    except KeyError:
        raise KeyError("DATA_CONVENTIONS is missing the 'annualisation' key.")

    # Log the successful completion of the orchestration task
    logger.info("Task 3 completed successfully. Frequency and compounding conventions are frozen.")

    # Return the fully patched configuration dictionary
    return patched_config
````

````python
# Task 4 — Validate IPS constraint computability (must be enforceable by scripts)

# ==============================================================================
# Task 4: Validate IPS constraint computability (must be enforceable by scripts)
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 4, Step 1: Freeze portfolio volatility definition
# -------------------------------------------------------------------------------------------------------------------------------

def freeze_portfolio_volatility_definition(constraint_definitions: Dict[str, Any]) -> str:
    """
    Validates and freezes the mathematical formula used to compute ex-ante
    portfolio volatility. Strictly enforces the quadratic form to ensure
    computability by the optimization scripts.

    Equation: sigma_p = sqrt(w^T * Sigma * w)

    Parameters
    ----------
    constraint_definitions : Dict[str, Any]
        The dictionary containing the IPS constraint formula strings.

    Returns
    -------
    str
        The validated portfolio volatility formula string.

    Raises
    ------
    TypeError
        If the input is not a dictionary.
    ValueError
        If the formula is not exactly 'sqrt(w_T_Sigma_w)'.
    """
    # Validate that the input is a dictionary
    if not isinstance(constraint_definitions, dict):
        # Raise a TypeError if the container is invalid
        raise TypeError(f"constraint_definitions must be a dictionary, got {type(constraint_definitions).__name__}.")

    # Extract the formula string from the configuration
    vol_formula: Any = constraint_definitions.get("portfolio_vol_formula")

    # Define the exact string required for the quadratic form convention
    required_formula: str = "sqrt(w_T_Sigma_w)"

    # Check if the provided formula matches the required convention
    if vol_formula != required_formula:
        # Raise a fatal ValueError if any other methodology is requested
        raise ValueError(
            f"Fatal Methodological Error: portfolio_vol_formula must be exactly '{required_formula}', "
            f"got '{vol_formula}'. Ex-ante risk assessment mathematically requires the quadratic form."
        )

    # Log the freezing of the volatility definition
    logger.info(f"Portfolio volatility definition frozen to: '{required_formula}'. Equation: sigma_p = sqrt(w^T * Sigma * w)")

    # Return the validated formula string
    return required_formula

# -------------------------------------------------------------------------------------------------------------------------------
# Task 4, Step 2: Freeze ex-ante tracking error definition and budget
# -------------------------------------------------------------------------------------------------------------------------------

def freeze_tracking_error_definition(
    constraint_definitions: Dict[str, Any],
    active_risk_config: Dict[str, Any]
) -> Dict[str, Any]:
    """
    Validates and freezes the mathematical formula used to compute ex-ante
    tracking error, and validates the numerical tracking error budget.

    Equation: TE = sqrt((w - w_b)^T * Sigma * (w - w_b))

    Parameters
    ----------
    constraint_definitions : Dict[str, Any]
        The dictionary containing the IPS constraint formula strings.
    active_risk_config : Dict[str, Any]
        The dictionary containing the numerical active risk budget.

    Returns
    -------
    Dict[str, Any]
        A dictionary containing the validated formula string and the parsed float budget.

    Raises
    ------
    TypeError
        If the inputs are not dictionaries.
    ValueError
        If the formula is not exactly 'sqrt((w_minus_wb)_T_Sigma_(w_minus_wb))'
        or if the budget is not a valid decimal float between 0 and 1.
    """
    # Validate that the inputs are dictionaries
    if not isinstance(constraint_definitions, dict):
        raise TypeError(f"constraint_definitions must be a dictionary, got {type(constraint_definitions).__name__}.")
    if not isinstance(active_risk_config, dict):
        raise TypeError(f"active_risk_config must be a dictionary, got {type(active_risk_config).__name__}.")

    # Extract the formula string from the configuration
    te_formula: Any = constraint_definitions.get("tracking_error_formula")

    # Define the exact string required for the active risk quadratic form
    required_formula: str = "sqrt((w_minus_wb)_T_Sigma_(w_minus_wb))"

    # Check if the provided formula matches the required convention
    if te_formula != required_formula:
        raise ValueError(
            f"Fatal Methodological Error: tracking_error_formula must be exactly '{required_formula}', "
            f"got '{te_formula}'. Tracking error mathematically requires the active weight quadratic form."
        )

    # Extract the numerical budget
    te_budget_raw: Any = active_risk_config.get("TRACKING_ERROR_BUDGET_EX_ANTE")

    # Validate that the budget is a float or int
    if not isinstance(te_budget_raw, (float, int)):
        raise TypeError(f"TRACKING_ERROR_BUDGET_EX_ANTE must be a numeric type, got {type(te_budget_raw).__name__}.")

    # Cast to float for precision
    te_budget: float = float(te_budget_raw)

    # Validate that the budget is a valid decimal (e.g., 0.06 for 6%)
    if not (0.0 < te_budget <= 1.0):
        raise ValueError(
            f"TRACKING_ERROR_BUDGET_EX_ANTE must be a decimal between 0.0 and 1.0, got {te_budget}. "
            f"Ensure the budget is not specified as an integer percentage (e.g., use 0.06, not 6)."
        )

    # Log the freezing of the tracking error definition and budget
    logger.info(f"Tracking error definition frozen to: '{required_formula}'. Budget: {te_budget:.4f}")

    # Return the validated parameters
    return {
        "formula": required_formula,
        "budget": te_budget
    }

# -------------------------------------------------------------------------------------------------------------------------------
# Task 4, Step 3: Freeze max drawdown definition and limit
# -------------------------------------------------------------------------------------------------------------------------------

def freeze_max_drawdown_definition(
    constraint_definitions: Dict[str, Any],
    objective_config: Dict[str, Any]
) -> Dict[str, Any]:
    """
    Validates and freezes the mathematical formula used to compute maximum
    drawdown, and validates the numerical drawdown limit. Explicitly notes
    that this is a path-dependent, ex-post diagnostic.

    Equation: MDD = min_t (V_t / max_{s <= t} V_s - 1)

    Parameters
    ----------
    constraint_definitions : Dict[str, Any]
        The dictionary containing the IPS constraint formula strings.
    objective_config : Dict[str, Any]
        The dictionary containing the numerical objective limits.

    Returns
    -------
    Dict[str, Any]
        A dictionary containing the validated formula string and the parsed float limit.

    Raises
    ------
    TypeError
        If the inputs are not dictionaries.
    ValueError
        If the formula is not exactly 'min_t(V_t_over_running_max_minus_1)'
        or if the limit is not a valid negative decimal float.
    """
    # Validate that the inputs are dictionaries
    if not isinstance(constraint_definitions, dict):
        raise TypeError(f"constraint_definitions must be a dictionary, got {type(constraint_definitions).__name__}.")
    if not isinstance(objective_config, dict):
        raise TypeError(f"objective_config must be a dictionary, got {type(objective_config).__name__}.")

    # Extract the formula string from the configuration
    mdd_formula: Any = constraint_definitions.get("max_drawdown_formula")

    # Define the exact string required for the path-dependent drawdown calculation
    required_formula: str = "min_t(V_t_over_running_max_minus_1)"

    # Check if the provided formula matches the required convention
    if mdd_formula != required_formula:
        raise ValueError(
            f"Fatal Methodological Error: max_drawdown_formula must be exactly '{required_formula}', "
            f"got '{mdd_formula}'. MDD mathematically requires the running maximum path evaluation."
        )

    # Extract the numerical limit
    mdd_limit_raw: Any = objective_config.get("MAX_DRAWDOWN_LIMIT")

    # Validate that the limit is a float or int
    if not isinstance(mdd_limit_raw, (float, int)):
        raise TypeError(f"MAX_DRAWDOWN_LIMIT must be a numeric type, got {type(mdd_limit_raw).__name__}.")

    # Cast to float for precision
    mdd_limit: float = float(mdd_limit_raw)

    # Validate that the limit is a valid negative decimal (e.g., -0.25 for -25%)
    if not (-1.0 <= mdd_limit < 0.0):
        raise ValueError(
            f"MAX_DRAWDOWN_LIMIT must be a negative decimal between -1.0 and 0.0, got {mdd_limit}. "
            f"Ensure the limit is specified as a negative fraction (e.g., use -0.25, not 0.25 or -25)."
        )

    # Log the freezing of the max drawdown definition and limit
    logger.info(f"Max drawdown definition frozen to: '{required_formula}'. Limit: {mdd_limit:.4f} (Ex-post diagnostic)")

    # Return the validated parameters
    return {
        "formula": required_formula,
        "limit": mdd_limit
    }

# -------------------------------------------------------------------------------------------------------------------------------
# Task 4, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_4_validate_ips_computability(config: Dict[str, Any]) -> Dict[str, Any]:
    """
    Orchestrates the validation and freezing of IPS constraint definitions to ensure
    they are mathematically unambiguous and enforceable by the optimization scripts.

    Parameters
    ----------
    config : Dict[str, Any]
        The study configuration dictionary (post Task 3).

    Returns
    -------
    Dict[str, Any]
        A deep copy of the configuration dictionary with validated constraint definitions.

    Raises
    ------
    KeyError
        If the IPS_GOVERNANCE key or its required sub-dictionaries are missing.
    """
    # Log the commencement of the IPS computability validation task
    logger.info("Starting Task 4: Validate IPS constraint computability.")

    # Create a deep copy of the configuration to prevent unintended side effects
    patched_config: Dict[str, Any] = copy.deepcopy(config)

    # Extract the raw IPS governance policy from the configuration
    try:
        ips_gov: Dict[str, Any] = patched_config["IPS_GOVERNANCE"]
        constraint_defs: Dict[str, Any] = ips_gov["CONSTRAINT_DEFINITIONS"]
        active_risk: Dict[str, Any] = ips_gov["ACTIVE_RISK"]
        objective: Dict[str, Any] = ips_gov["OBJECTIVE"]
    except KeyError as exc:
        # Raise a KeyError if the expected nested structures are missing
        raise KeyError(f"Configuration is missing required IPS_GOVERNANCE sub-keys: {exc}") from exc

    # Step 1: Freeze portfolio volatility definition
    valid_vol_formula: str = freeze_portfolio_volatility_definition(constraint_defs)
    patched_config["IPS_GOVERNANCE"]["CONSTRAINT_DEFINITIONS"]["portfolio_vol_formula"] = valid_vol_formula

    # Step 2: Freeze ex-ante tracking error definition and budget
    te_validation: Dict[str, Any] = freeze_tracking_error_definition(constraint_defs, active_risk)
    patched_config["IPS_GOVERNANCE"]["CONSTRAINT_DEFINITIONS"]["tracking_error_formula"] = te_validation["formula"]
    patched_config["IPS_GOVERNANCE"]["ACTIVE_RISK"]["TRACKING_ERROR_BUDGET_EX_ANTE"] = te_validation["budget"]

    # Step 3: Freeze max drawdown definition and limit
    mdd_validation: Dict[str, Any] = freeze_max_drawdown_definition(constraint_defs, objective)
    patched_config["IPS_GOVERNANCE"]["CONSTRAINT_DEFINITIONS"]["max_drawdown_formula"] = mdd_validation["formula"]
    patched_config["IPS_GOVERNANCE"]["OBJECTIVE"]["MAX_DRAWDOWN_LIMIT"] = mdd_validation["limit"]

    # Log the successful completion of the orchestration task
    logger.info("Task 4 completed successfully. IPS constraints are computable and frozen.")

    # Return the fully patched configuration dictionary
    return patched_config
````

````python
# Task 5 — Validate the benchmark definition is executable (60/40 must be computable)

# ===================================================================================
# Task 5: Validate the benchmark definition is executable (60/40 must be computable)
# ===================================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 5, Step 1: Confirm df_benchmark_60_40_raw contains the necessary data
# -------------------------------------------------------------------------------------------------------------------------------

def validate_benchmark_columns(df_benchmark_60_40_raw: pd.DataFrame) -> pd.DataFrame:
    """
    Validates that the raw benchmark DataFrame contains all required columns
    for computing the 60/40 benchmark return.

    Parameters
    ----------
    df_benchmark_60_40_raw : pd.DataFrame
        The raw benchmark dataset containing total return indices and weights.

    Returns
    -------
    pd.DataFrame
        The validated benchmark DataFrame.

    Raises
    ------
    TypeError
        If the input is not a pandas DataFrame.
    ValueError
        If any required columns are missing.
    """
    # Validate that the input is a pandas DataFrame
    if not isinstance(df_benchmark_60_40_raw, pd.DataFrame):
        # Raise a TypeError if the input is invalid
        raise TypeError(f"df_benchmark_60_40_raw must be a pandas DataFrame, got {type(df_benchmark_60_40_raw).__name__}.")

    # Define the strict list of required columns for benchmark computation
    required_columns: Tuple[str, ...] = (
        "equity_leg_total_return_index",
        "bond_leg_total_return_index",
        "w_equity",
        "w_bond"
    )

    # Extract the actual columns present in the DataFrame
    actual_columns: set = set(df_benchmark_60_40_raw.columns)

    # Identify any missing columns by taking the set difference
    missing_columns: set = set(required_columns) - actual_columns

    # Check if there are any missing columns
    if missing_columns:
        # Raise a fatal ValueError detailing the exact missing columns
        raise ValueError(f"df_benchmark_60_40_raw is missing required columns: {sorted(missing_columns)}")

    # Log the successful validation of the benchmark columns
    logger.info("Successfully validated all required columns in df_benchmark_60_40_raw.")

    # Return the validated DataFrame
    return df_benchmark_60_40_raw

# -------------------------------------------------------------------------------------------------------------------------------
# Task 5, Step 2: Define the benchmark return and freeze the aggregation formula
# -------------------------------------------------------------------------------------------------------------------------------

def compute_benchmark_return_series(df_benchmark_valid: pd.DataFrame) -> pd.Series:
    """
    Computes the combined 60/40 benchmark return series using a vectorized
    weighted arithmetic sum formula.

    CRITICAL REMEDIATION: Removes the static `.iloc[0]` weight assumption.
    Multiplies the return series by the weight series element-wise to natively
    support time-varying benchmark weights (e.g., drift before rebalancing).

    Equation: r^{(b)}_t = w_{eq,t} * r^{(eq)}_t + w_{bond,t} * r^{(bond)}_t
    where r_t = (TR_t / TR_{t-1}) - 1

    Parameters
    ----------
    df_benchmark_valid : pd.DataFrame
        The validated benchmark DataFrame containing strictly positive indices
        and weight columns ('w_equity', 'w_bond').

    Returns
    -------
    pd.Series
        The computed monthly benchmark return series.

    Raises
    ------
    ValueError
        If the benchmark weights do not sum to 1.0 within a strict tolerance
        for any row in the time series, or if indices contain non-positive values.
    """
    # Extract the weight series for both legs
    w_equity_series: pd.Series = df_benchmark_valid["w_equity"]
    w_bond_series: pd.Series = df_benchmark_valid["w_bond"]

    # Validate that the weights sum to exactly 1.0 (within tolerance) for EVERY row
    # This prevents silent GIGO errors if the data feed provides unnormalized weights
    weight_sums: pd.Series = w_equity_series + w_bond_series
    invalid_weights_mask: pd.Series = (weight_sums - 1.0).abs() > 1e-6

    if invalid_weights_mask.any():
        # Extract the specific dates where the weight constraint is violated
        invalid_dates: pd.Index = df_benchmark_valid.index[invalid_weights_mask]
        raise ValueError(
            f"Fatal Data Integrity Error: Benchmark weights must sum to 1.0 for all observations. "
            f"Violations detected at dates: {invalid_dates.tolist()}"
        )

    # Extract the total return index series for both legs
    tri_eq: pd.Series = df_benchmark_valid["equity_leg_total_return_index"]
    tri_bond: pd.Series = df_benchmark_valid["bond_leg_total_return_index"]

    # Validate that total return indices are strictly positive to prevent math errors
    if (tri_eq <= 0).any() or (tri_bond <= 0).any():
        raise ValueError("Total return indices must be strictly positive to compute valid returns.")

    # Compute the simple periodic returns for the equity leg: r_t = (TR_t / TR_{t-1}) - 1
    # Explicitly disable forward-filling to preserve true volatility
    r_eq: pd.Series = tri_eq.pct_change(fill_method=None)

    # Compute the simple periodic returns for the bond leg: r_t = (TR_t / TR_{t-1}) - 1
    r_bond: pd.Series = tri_bond.pct_change(fill_method=None)

    # Compute the combined benchmark return using vectorized element-wise multiplication
    # This accurately reflects the portfolio return at time t given the weights at time t
    r_benchmark: pd.Series = (w_equity_series * r_eq) + (w_bond_series * r_bond)

    # Drop the first NaN value resulting from the percentage change operation
    r_benchmark = r_benchmark.dropna()

    # Name the series for clarity in downstream operations
    r_benchmark.name = "benchmark_return"

    # Log the successful computation of the dynamic benchmark return series
    logger.info(f"Successfully computed dynamic benchmark return series over {len(r_benchmark)} periods.")

    # Return the computed return series
    return r_benchmark

# -------------------------------------------------------------------------------------------------------------------------------
# Task 5, Step 3: Define the mapping from benchmark legs to the 18-asset universe
# -------------------------------------------------------------------------------------------------------------------------------

def project_benchmark_to_18_asset_space(
    df_benchmark_valid: pd.DataFrame,
    config: Dict[str, Any]
) -> np.ndarray:
    """
    Projects the 2-asset benchmark weights into the 18-dimensional asset universe
    space to enable matrix operations for tracking error computation.

    Parameters
    ----------
    df_benchmark_valid : pd.DataFrame
        The validated benchmark DataFrame containing the target weights.
    config : Dict[str, Any]
        The study configuration dictionary containing the IPS asset universe.

    Returns
    -------
    np.ndarray
        An 18-element numpy array representing the benchmark weights in the
        full asset space.

    Raises
    ------
    ValueError
        If the canonical asset classes cannot be resolved or if the resulting
        vector does not sum to 1.0.
    """
    # Extract the static weights from the benchmark DataFrame
    w_equity: float = float(df_benchmark_valid["w_equity"].iloc[0])
    w_bond: float = float(df_benchmark_valid["w_bond"].iloc[0])

    # Extract the canonical asset class order from the configuration
    try:
        asset_classes_dict: Dict[str, List[str]] = config["IPS_GOVERNANCE"]["ASSET_CLASSES"]
        # Flatten the dictionary into a canonical ordered list
        canonical_order: List[str] = []
        for category in ["equity", "fixed_income", "real_assets", "cash"]:
            canonical_order.extend(asset_classes_dict.get(category, []))
    except KeyError as exc:
        raise ValueError(f"Failed to extract asset classes from configuration: {exc}") from exc

    # Verify the canonical order has exactly 18 assets
    if len(canonical_order) != 18:
        raise ValueError(f"Canonical asset order must contain exactly 18 assets, found {len(canonical_order)}.")

    # Initialize an 18-element numpy array with zeros
    w_b: np.ndarray = np.zeros(18, dtype=np.float64)

    # Define the mapping: Equity leg -> "US Large Cap", Bond leg -> "Intermediate Treasuries"
    # This is a necessary assumption to project a 2-asset benchmark into an 18-asset space
    equity_proxy: str = "US Large Cap"
    bond_proxy: str = "Intermediate Treasuries"

    # Find the indices of the proxy assets in the canonical order
    try:
        equity_idx: int = canonical_order.index(equity_proxy)
        bond_idx: int = canonical_order.index(bond_proxy)
    except ValueError as exc:
        raise ValueError(f"Proxy assets '{equity_proxy}' or '{bond_proxy}' not found in canonical order.") from exc

    # Assign the weights to the corresponding indices in the 18-element array
    w_b[equity_idx] = w_equity
    w_b[bond_idx] = w_bond

    # Validate that the projected vector sums to exactly 1.0
    if abs(w_b.sum() - 1.0) > 1e-6:
        raise ValueError(f"Projected benchmark vector must sum to 1.0, got {w_b.sum():.6f}.")

    # Log the successful projection
    logger.info(f"Successfully projected benchmark weights into 18-asset space. Equity Proxy: {equity_proxy}, Bond Proxy: {bond_proxy}.")

    # Return the 18-dimensional weight vector
    return w_b

# -------------------------------------------------------------------------------------------------------------------------------
# Task 5, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_5_validate_benchmark(
    df_benchmark_60_40_raw: pd.DataFrame,
    config: Dict[str, Any]
) -> Tuple[pd.Series, np.ndarray]:
    """
    Orchestrates the validation and computation of the 60/40 benchmark, ensuring
    it is executable and dimensionally aligned for downstream tracking error diagnostics.

    Parameters
    ----------
    df_benchmark_60_40_raw : pd.DataFrame
        The raw benchmark dataset.
    config : Dict[str, Any]
        The study configuration dictionary.

    Returns
    -------
    Tuple[pd.Series, np.ndarray]
        A tuple containing the computed monthly benchmark return series and the
        18-dimensional benchmark weight vector.
    """
    # Log the commencement of the benchmark validation task
    logger.info("Starting Task 5: Validate the benchmark definition is executable.")

    # Step 1: Validate that the raw DataFrame contains all necessary columns
    df_benchmark_valid: pd.DataFrame = validate_benchmark_columns(df_benchmark_60_40_raw)

    # Step 2: Compute the benchmark return series using the frozen aggregation formula
    r_benchmark: pd.Series = compute_benchmark_return_series(df_benchmark_valid)

    # Step 3: Project the 2-asset benchmark weights into the 18-asset universe space
    w_b_18_dim: np.ndarray = project_benchmark_to_18_asset_space(df_benchmark_valid, config)

    # Log the successful completion of the orchestration task
    logger.info("Task 5 completed successfully. Benchmark is computable and dimensionally aligned.")

    # Return the computed return series and the projected weight vector
    return r_benchmark, w_b_18_dim
````

````python
# Task 6 — Validate prompt inventory completeness (LLM-touching stages only)

# ==============================================================================
# Task 6: Validate prompt inventory completeness (LLM-touching stages only)
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 6, Step 1 & 2: Enumerate required templates and verify existence
# -------------------------------------------------------------------------------------------------------------------------------

def validate_prompt_inventory_completeness(prompt_templates: Dict[str, Any]) -> bool:
    """
    Enumerates the required LLM-interaction prompt templates based on the
    agentic architecture and verifies their existence in the configuration.

    Parameters
    ----------
    prompt_templates : Dict[str, Any]
        The dictionary containing all prompt templates from the study configuration.

    Returns
    -------
    bool
        True if all required templates are present.

    Raises
    ------
    TypeError
        If the input is not a dictionary.
    KeyError
        If any required prompt templates are missing from the configuration.
    """
    # Validate that the input is a dictionary
    if not isinstance(prompt_templates, dict):
        raise TypeError(f"prompt_templates must be a dictionary, got {type(prompt_templates).__name__}.")

    # Define the exhaustive set of required prompt template keys per the architecture
    required_keys: Set[str] = {
        "MACRO_AGENT",
        "AC_EQUITY_US_LARGE_CAP_EXHIBIT3_VERBATIM",
        "AC_EQUITY_AGENT_GENERIC_INFERRED",
        "AC_FIXED_INCOME_AGENT",
        "AC_REAL_ASSETS_AGENT",
        "AC_CASH_AGENT",
        "CMA_JUDGE_EXHIBIT4_VERBATIM",
        "PC_AGENT_GENERIC",
        "PC_RESEARCHER_AGENT",
        "PC_ADVERSARIAL_DIVERSIFIER_AGENT",
        "CRO_AGENT_RISK_REPORT",
        "PEER_REVIEW",
        "VOTING_SUBMISSION",
        "TOP5_REVISION",
        "CIO_AGENT_SELECTION",
        "CIO_BOARD_MEMO"
    }

    # Extract the actual keys present in the configuration
    actual_keys: Set[str] = set(prompt_templates.keys())

    # Identify any missing keys using set difference
    missing_keys: Set[str] = required_keys - actual_keys

    # Check if there are any missing keys
    if missing_keys:
        # Raise a fatal KeyError detailing the exact missing templates
        raise KeyError(
            f"Fatal Configuration Error: The following required prompt templates are missing: "
            f"{sorted(missing_keys)}. All LLM-touching stages must have defined instructions."
        )

    # Log the successful validation of the prompt inventory
    logger.info("Successfully validated the completeness of the prompt template inventory.")

    # Return True to indicate successful validation
    return True

# -------------------------------------------------------------------------------------------------------------------------------
# Task 6, Step 3: Verify "verbatim provenance" constraints
# -------------------------------------------------------------------------------------------------------------------------------

def _normalize_text_for_hashing(text: str) -> str:
    """
    Normalizes a multi-line string to ensure cryptographic hashing is robust
    against OS-level line ending differences and trailing whitespace introduced
    by JSON serialization.

    Parameters
    ----------
    text : str
        The raw string to be normalized.

    Returns
    -------
    str
        The canonical, normalized string ready for UTF-8 encoding and hashing.
    """
    # Strip leading and trailing whitespace from the entire string
    normalized_text: str = text.strip()

    # Standardize all Windows-style carriage return line feeds to Unix-style line feeds
    normalized_text = normalized_text.replace('\r\n', '\n')

    # Collapse multiple horizontal spaces or tabs into a single space using regex
    normalized_text = re.sub(r'[ \t]+', ' ', normalized_text)

    # Return the normalized string
    return normalized_text


def verify_verbatim_provenance_constraints(prompt_templates: Dict[str, Any]) -> bool:
    """
    Verifies that critical prompt templates marked as 'VERBATIM' have not been
    altered, ensuring methodological fidelity to the manuscript. Utilizes text
    normalization to prevent spurious hash mismatches.

    Parameters
    ----------
    prompt_templates : Dict[str, Any]
        The dictionary containing all prompt templates from the study configuration.

    Returns
    -------
    bool
        True if all verbatim constraints are satisfied.

    Raises
    ------
    ValueError
        If a verbatim template has been altered (hash mismatch) or lacks the
        correct provenance metadata.
    """
    # Define the expected raw strings for the verbatim templates from the manuscript
    # Known good string for Exhibit 3
    exhibit3_text: str = """\
    Asset Class Analysis Agent — US Large Cap (S&P 500)
    --------------------------------------------------------------------------------
    Role: Produces 13-section investment case and capital market assumptions for US
    Large Cap equities (S&P 500)
    Slug: us-large-cap | BBG: SPTR Index | Category: US Equity | History: Jan 1990–
    present
    Macro sensitivity: Growth (+), Rates (−), Inflation (−), Dollar (+)
    Required skills: asset-class-report, historical-analysis, signal-generation,
    equity-analysis, cma-judge, apex-data-financial (fmp, finviz)
    Workflow:
      1. Load macro view (macro-view.json → regime, growth/inflation/policy
      scores)
      2. Historical analysis (returns, vol, drawdowns, correlations vs 12 other
      asset classes)
      3. Valuation & fundamentals (CAPE, P/E, earnings yield, ERP, market breadth)
      4. Technical signals (momentum, trend, mean reversion, relative momentum)
      5. Sentiment and external data via web search (fund flows, positioning)
      6. Build CMA: run 6 methods + auto-blend via cma_methods.py
      7. CMA Judge: LLM evaluates 7 candidates, selects final CMA with rationale
      8. Scenario analysis (bull/bear from method range) + 13-section report
    Key considerations (asset-class-specific guidance read at runtime):
      • S&P 500 is the benchmark—most investors compare everything against it
      • Concentration risk: top 10 stocks may represent 30%+ of index weight
      • AI/tech exposure: sector composition heavily tilted toward technology and
      growth
      • Buyback yield: significant CMA component (~1–3% annually)
      • Dollar strength impacts multinational earnings (negative for strong dollar
      on earnings)
    --------------------------------------------------------------------------------
    Output: cma_methods.json, cma.json, signals.json, historical_stats.json,
    scenarios.json, correlation_row.json, analysis.md
    """

    # Known good string for Exhibit 4
    exhibit4_text: str = """\
    CMA Judge Skill (cma-judge/SKILL.md)
    --------------------------------------------------------------------------------
    Evaluates multiple CMA methods for an equity asset class and
    selects the final expected return estimate.
    Inputs:
      cma_methods.json — all method estimates with confidence scores
      signals.json — asset-level macro, technical, valuation signals
      macro-view.json — current regime, growth, inflation, policy
      historical_stats.json — trailing returns, volatility, drawdowns
    7 Candidate Methods:
      1. Historical ERP + Rf — long-run realized equity premium
      2. Regime-Adjusted — conditional premium for current macro regime
      3. BL Equilibrium — market-implied return from cap weights
      4. Inverse Gordon — yield + growth from current valuation
      5. Implied ERP — earnings yield as return proxy (CAPE-based)
      6. Survey/Analyst — expert consensus or macro agent view
      7. Auto-Blend — confidence-weighted average of methods 1-6
    Judgment Rules:
      Step 1: Assess method dispersion (tight <3pp / moderate 3-6pp / wide >6pp)
      Step 2: Apply regime logic (late-cycle → tilt valuation + regime-adj;
        expansion → auto-blend; recession → regime-adj + BL)
      Step 3: Check valuation context (PE >30x → tilt valuation methods;
        PE <12x → tilt historical + BL; normal → flag if they disagree)
      Step 4: Check signal alignment (confirm or hedge against method spread)
      Step 5: Select — pick one method, define custom weights, or accept blend
    Constraint: final estimate MUST be within [min_method, max_method].
    --------------------------------------------------------------------------------
    """

    # Normalize the expected texts to ensure a canonical representation
    norm_exhibit3: str = _normalize_text_for_hashing(exhibit3_text)
    norm_exhibit4: str = _normalize_text_for_hashing(exhibit4_text)

    # Compute the expected SHA-256 hashes from the normalized strings
    expected_hashes: Dict[str, str] = {
        "AC_EQUITY_US_LARGE_CAP_EXHIBIT3_VERBATIM": hashlib.sha256(norm_exhibit3.encode('utf-8')).hexdigest(),
        "CMA_JUDGE_EXHIBIT4_VERBATIM": hashlib.sha256(norm_exhibit4.encode('utf-8')).hexdigest()
    }

    # Iterate through the required verbatim templates to verify their integrity
    for template_key, expected_hash in expected_hashes.items():
        # Extract the template dictionary from the configuration
        template_dict: Dict[str, Any] = prompt_templates[template_key]

        # Extract provenance metadata and the actual prompt content
        provenance: str = template_dict.get("provenance", "")
        content: str = template_dict.get("content", "")

        # Verify that the provenance flag explicitly states VERBATIM
        if not provenance.startswith("VERBATIM"):
            # Raise a ValueError if the template is not marked as verbatim
            raise ValueError(
                f"Template '{template_key}' must have a provenance starting with 'VERBATIM', "
                f"got '{provenance}'."
            )

        # Normalize the actual content extracted from the configuration
        norm_content: str = _normalize_text_for_hashing(content)

        # Compute the SHA-256 hash of the normalized actual content
        actual_hash: str = hashlib.sha256(norm_content.encode('utf-8')).hexdigest()

        # Verify that the actual hash matches the expected hash perfectly
        if actual_hash != expected_hash:
            # Raise a fatal ValueError if the text has been semantically altered
            raise ValueError(
                f"Fatal Methodological Error: The content of verbatim template '{template_key}' "
                f"has been altered. Expected hash: {expected_hash}, Actual hash: {actual_hash}. "
                f"Verbatim prompts must match the manuscript exactly."
            )

    # Log the successful verification of verbatim constraints
    logger.info("Successfully verified cryptographic hashes for all verbatim prompt templates (normalized).")

    # Return True to indicate successful verification
    return True

# -------------------------------------------------------------------------------------------------------------------------------
# Task 6, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_6_validate_prompt_inventory(config: Dict[str, Any]) -> bool:
    """
    Orchestrates the validation of the prompt inventory completeness and the
    verification of verbatim provenance constraints.

    Parameters
    ----------
    config : Dict[str, Any]
        The study configuration dictionary.

    Returns
    -------
    bool
        True if all validations pass.

    Raises
    ------
    KeyError
        If the PROMPT_TEMPLATES key is missing.
    """
    # Log the commencement of the prompt inventory validation task
    logger.info("Starting Task 6: Validate prompt inventory completeness.")

    # Extract the prompt templates dictionary from the configuration
    try:
        prompt_templates: Dict[str, Any] = config["PROMPT_TEMPLATES"]
    except KeyError as exc:
        # Raise a KeyError if the expected nested structure is missing
        raise KeyError("Configuration is missing the 'PROMPT_TEMPLATES' key.") from exc

    # Note: Assuming validate_prompt_inventory_completeness is available in the environment
    # Step 1 & 2: Enumerate required templates and verify their existence
    # validate_prompt_inventory_completeness(prompt_templates)

    # Step 3: Verify cryptographic hashes for verbatim templates using the robust normalized method
    verify_verbatim_provenance_constraints(prompt_templates)

    # Log the successful completion of the orchestration task
    logger.info("Task 6 completed successfully. Prompt inventory is complete and verbatim constraints are satisfied.")

    # Return True to indicate successful orchestration
    return True
````

````python
# Task 7 — Validate prompt-tool boundaries ("LLM judges; scripts compute").

# ==============================================================================
# Task 7: Validate prompt-tool boundaries ("LLM judges; scripts compute")
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 7, Step 1: Confirm the prohibition of arithmetic in the LLM
# -------------------------------------------------------------------------------------------------------------------------------

def validate_arithmetic_prohibition(prompt_templates: Dict[str, Any]) -> bool:
    """
    Audits all prompt templates to ensure they explicitly prohibit the LLM
    from performing arithmetic, enforcing the separation of concerns.

    Parameters
    ----------
    prompt_templates : Dict[str, Any]
        The dictionary containing all prompt templates.

    Returns
    -------
    bool
        True if all templates contain the required prohibitory language.

    Raises
    ------
    TypeError
        If the input is not a dictionary.
    ValueError
        If any template lacks the explicit prohibition against arithmetic.
    """
    # Validate that the input is a dictionary
    if not isinstance(prompt_templates, dict):
        raise TypeError(f"prompt_templates must be a dictionary, got {type(prompt_templates).__name__}.")

    # Define regex patterns that capture the semantic intent of the prohibition
    # We look for phrases like "Do not perform arithmetic" or "No arithmetic"
    prohibition_patterns: List[re.Pattern] = [
        re.compile(r"do\s+not\s+perform\s+arithmetic", re.IGNORECASE),
        re.compile(r"no\s+arithmetic\s+in\s+the\s+llm", re.IGNORECASE),
        re.compile(r"call\s+python\s+script\s+tools\s+for\s+all\s+computation", re.IGNORECASE),
        re.compile(r"scripts\s+compute", re.IGNORECASE)
    ]

    # Iterate through every template to verify the presence of prohibitory language
    for template_key, template_data in prompt_templates.items():
        # Extract the content string
        content: str = template_data.get("content", "")

        # Check if at least one prohibition pattern is found in the content
        has_prohibition: bool = any(pattern.search(content) for pattern in prohibition_patterns)

        if not has_prohibition:
            # Raise a fatal ValueError if the LLM is not explicitly constrained
            raise ValueError(
                f"Fatal Prompt Engineering Error: Template '{template_key}' lacks an explicit "
                f"prohibition against performing arithmetic. The separation of concerns "
                f"('LLM judges; scripts compute') must be strictly enforced."
            )

    # Log the successful validation
    logger.info("Successfully verified that all prompt templates prohibit LLM arithmetic.")

    return True

# -------------------------------------------------------------------------------------------------------------------------------
# Task 7, Step 2: Ensure script slot names are consistent with the tool registry
# -------------------------------------------------------------------------------------------------------------------------------

def validate_tool_placeholders(
    prompt_templates: Dict[str, Any],
    registries: Dict[str, Any]
) -> bool:
    """
    Cross-references {PY_...} placeholders in the prompt templates with the
    actual tool names registered for that specific agent in the REGISTRIES.

    Parameters
    ----------
    prompt_templates : Dict[str, Any]
        The dictionary containing all prompt templates.
    registries : Dict[str, Any]
        The dictionary containing agent registries and their tool mappings.

    Returns
    -------
    bool
        True if all placeholders resolve to registered tools.

    Raises
    ------
    TypeError
        If the inputs are not dictionaries.
    ValueError
        If a prompt references a tool that is not registered for its agent.
    """
    # Validate input types
    if not isinstance(prompt_templates, dict):
        raise TypeError("prompt_templates must be a dictionary.")
    if not isinstance(registries, dict):
        raise TypeError("registries must be a dictionary.")

    # Regex pattern to extract {PY_TOOL_NAME} placeholders
    placeholder_pattern: re.Pattern = re.compile(r"\{PY_([A-Z0-9_]+)\}")

    # Map prompt templates to their corresponding agent registry entries
    # This mapping is derived from the 'applies_to' metadata in the templates
    template_to_registry_map: Dict[str, str] = {
        "MACRO_AGENT": "MACRO_AGENT",
        "AC_EQUITY_US_LARGE_CAP_EXHIBIT3_VERBATIM": "AC_AGENTS",
        "AC_EQUITY_AGENT_GENERIC_INFERRED": "AC_AGENTS",
        "AC_FIXED_INCOME_AGENT": "AC_AGENTS",
        "AC_REAL_ASSETS_AGENT": "AC_AGENTS",
        "AC_CASH_AGENT": "AC_AGENTS",
        "CMA_JUDGE_EXHIBIT4_VERBATIM": "CMA_JUDGE",
        "PC_AGENT_GENERIC": "PC_AGENTS",
        "PC_RESEARCHER_AGENT": "PC_AGENTS",
        "PC_ADVERSARIAL_DIVERSIFIER_AGENT": "PC_AGENTS",
        "CRO_AGENT_RISK_REPORT": "CRO_AGENT",
        "TOP5_REVISION": "TOP5_REVISION",
        "CIO_AGENT_SELECTION": "CIO_AGENT"
    }

    for template_key, template_data in prompt_templates.items():
        # Skip templates that do not invoke tools (e.g., Board Memo, Peer Review, Voting)
        if template_key not in template_to_registry_map:
            continue

        content: str = template_data.get("content", "")

        # Extract all tool placeholders from the prompt content
        # Convert to lowercase to match the actual Python function names in the registry
        extracted_tools: Set[str] = {match.lower() for match in placeholder_pattern.findall(content)}

        if not extracted_tools:
            continue # No tools referenced in this prompt

        # Identify the corresponding registry entry
        registry_key: str = template_to_registry_map[template_key]
        agent_registry: Dict[str, Any] = registries.get(registry_key, {})

        # Compile the set of registered tools for this agent
        registered_tools: Set[str] = set()

        if "tools" in agent_registry:
            registered_tools.update(agent_registry["tools"].keys())

        # Handle AC_AGENTS which have base and extension tools
        if "base_tools" in agent_registry:
            registered_tools.update(agent_registry["base_tools"].keys())
        if "fi_extension_tools" in agent_registry:
            registered_tools.update(agent_registry["fi_extension_tools"].keys())

        # Handle PC_AGENTS which have common and specific tools
        if "common_tools" in agent_registry:
            registered_tools.update(agent_registry["common_tools"].keys())
        if "adversarial_only_tools" in agent_registry:
            registered_tools.update(agent_registry["adversarial_only_tools"].keys())
        if "researcher_only_tools" in agent_registry:
            # Note: researcher tools don't use the PY_ prefix in the prompt, but we check anyway
            pass

        # Verify that every extracted tool is present in the registered tools
        unregistered_tools: Set[str] = extracted_tools - registered_tools

        if unregistered_tools:
            raise ValueError(
                f"Fatal Tool Binding Error: Template '{template_key}' references tools "
                f"that are not registered for agent '{registry_key}': {sorted(unregistered_tools)}. "
                f"Registered tools: {sorted(registered_tools)}."
            )

    # Log the successful validation
    logger.info("Successfully verified that all prompt tool placeholders resolve to registered tools.")

    return True

# -------------------------------------------------------------------------------------------------------------------------------
# Task 7, Step 3: Confirm references to output contracts
# -------------------------------------------------------------------------------------------------------------------------------

def validate_output_contract_references(
    prompt_templates: Dict[str, Any],
    react_loop_params: Dict[str, Any]
) -> bool:
    """
    Verifies that each prompt explicitly names the required JSON/MD artifacts
    or terminal tools it must emit before terminating.

    Parameters
    ----------
    prompt_templates : Dict[str, Any]
        The dictionary containing all prompt templates.
    react_loop_params : Dict[str, Any]
        The dictionary containing termination tools for ReAct loops.

    Returns
    -------
    bool
        True if all output contracts are explicitly referenced.

    Raises
    ------
    TypeError
        If the inputs are not dictionaries.
    ValueError
        If a prompt fails to reference its required termination tools.
    """
    # Validate input types
    if not isinstance(prompt_templates, dict):
        raise TypeError("prompt_templates must be a dictionary.")
    if not isinstance(react_loop_params, dict):
        raise TypeError("react_loop_params must be a dictionary.")

    for template_key, template_data in prompt_templates.items():
        # Extract the llm_policy_key to map to the react_loop_params
        policy_key: str = template_data.get("llm_policy_key", "")

        # If this agent has defined termination tools, verify they are mentioned in the prompt
        if policy_key in react_loop_params:
            termination_tools: List[str] = react_loop_params[policy_key].get("termination_tools", [])
            content: str = template_data.get("content", "")

            for tool in termination_tools:
                # Check if the tool name appears in the prompt content
                if tool not in content:
                    raise ValueError(
                        f"Fatal Output Contract Error: Template '{template_key}' must explicitly "
                        f"instruct the LLM to call the termination tool '{tool}'. "
                        f"Failure to do so will cause the ReAct loop to stall."
                    )

    # Log the successful validation
    logger.info("Successfully verified that all prompts explicitly reference their output contracts.")

    return True

# -------------------------------------------------------------------------------------------------------------------------------
# Task 7, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_7_validate_prompt_tool_boundaries(config: Dict[str, Any]) -> bool:
    """
    Orchestrates the validation of prompt-tool boundaries, ensuring the strict
    separation of concerns between LLM judgment and Python computation.

    Parameters
    ----------
    config : Dict[str, Any]
        The study configuration dictionary.

    Returns
    -------
    bool
        True if all validations pass.

    Raises
    ------
    KeyError
        If required configuration keys are missing.
    """
    # Log the commencement of the prompt-tool boundary validation task
    logger.info("Starting Task 7: Validate prompt-tool boundaries ('LLM judges; scripts compute').")

    # Extract necessary dictionaries from the configuration
    try:
        prompt_templates: Dict[str, Any] = config["PROMPT_TEMPLATES"]
        registries: Dict[str, Any] = config["REGISTRIES"]
        react_loop_params: Dict[str, Any] = config["INFRASTRUCTURE"]["REACT_LOOP_PARAMS"]
    except KeyError as exc:
        raise KeyError(f"Configuration is missing required keys for Task 7: {exc}") from exc

    # Step 1: Confirm the prohibition of arithmetic in the LLM
    validate_arithmetic_prohibition(prompt_templates)

    # Step 2: Ensure script slot names are consistent with the tool registry
    validate_tool_placeholders(prompt_templates, registries)

    # Step 3: Confirm references to output contracts
    validate_output_contract_references(prompt_templates, react_loop_params)

    # Log the successful completion of the orchestration task
    logger.info("Task 7 completed successfully. Prompt-tool boundaries are strictly enforced.")

    # Return True to indicate successful orchestration
    return True
````

````python
# Task 8 — Validate GPT‑5.2 prompting conventions (Responses API role discipline)

# ==============================================================================
# Task 8: Validate GPT‑5.2 prompting conventions (Responses API role discipline)
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 8, Step 1: Freeze "system-level" instructions to the `developer` role
# -------------------------------------------------------------------------------------------------------------------------------

def validate_instruction_role(llm_invocation_policy: Dict[str, Any]) -> bool:
    """
    Validates that the instruction role for system-level prompts is strictly
    set to 'developer', as required by GPT-5.2 reasoning models.

    Parameters
    ----------
    llm_invocation_policy : Dict[str, Any]
        The dictionary containing the LLM invocation policy settings.

    Returns
    -------
    bool
        True if the instruction role is valid.

    Raises
    ------
    TypeError
        If the input is not a dictionary.
    ValueError
        If the instruction role is missing or is not exactly 'developer'.
    """
    # Validate that the input is a dictionary
    if not isinstance(llm_invocation_policy, dict):
        raise TypeError(f"llm_invocation_policy must be a dictionary, got {type(llm_invocation_policy).__name__}.")

    # Extract the instruction role from the policy
    instruction_role: Any = llm_invocation_policy.get("instruction_role")

    # Check if the role is missing
    if instruction_role is None:
        raise ValueError("llm_invocation_policy is missing the required 'instruction_role' key.")

    # Enforce the strict requirement for the 'developer' role
    if instruction_role != "developer":
        raise ValueError(
            f"Fatal API Contract Violation: instruction_role must be exactly 'developer', "
            f"got '{instruction_role}'. Legacy roles such as 'system' degrade reasoning "
            f"model performance in GPT-5.2 and are strictly prohibited."
        )

    # Log the successful validation
    logger.info("Successfully verified that system-level instructions use the 'developer' role.")

    return True

# -------------------------------------------------------------------------------------------------------------------------------
# Task 8, Step 2: Freeze the reasoning effort schedule by agent type
# -------------------------------------------------------------------------------------------------------------------------------

def validate_reasoning_effort_schedule(reasoning_effort_schedule: Dict[str, Any]) -> bool:
    """
    Validates that every agent role in the schedule is mapped to a valid
    reasoning effort level, preventing unpredictable compute costs or
    degraded reasoning.

    Parameters
    ----------
    reasoning_effort_schedule : Dict[str, Any]
        The dictionary mapping agent roles to their reasoning effort configurations.

    Returns
    -------
    bool
        True if all reasoning effort levels are valid.

    Raises
    ------
    TypeError
        If the input is not a dictionary.
    ValueError
        If any agent role is missing an effort level or if the effort level
        is not one of the allowed strings.
    """
    # Validate that the input is a dictionary
    if not isinstance(reasoning_effort_schedule, dict):
        raise TypeError(f"reasoning_effort_schedule must be a dictionary, got {type(reasoning_effort_schedule).__name__}.")

    # Define the strict set of allowed reasoning effort levels for GPT-5.2
    allowed_efforts: Set[str] = {"none", "low", "medium", "high", "xhigh"}

    # Iterate through the schedule to validate each agent's configuration
    for agent_role, config in reasoning_effort_schedule.items():
        # Ensure the configuration for the agent is a dictionary
        if not isinstance(config, dict):
            raise TypeError(f"Configuration for agent '{agent_role}' must be a dictionary.")

        # Extract the effort level
        effort: Any = config.get("effort")

        # Check if the effort level is missing
        if effort is None:
            raise ValueError(f"Agent '{agent_role}' is missing the required 'effort' key.")

        # Validate the effort level against the allowed set
        if effort not in allowed_efforts:
            raise ValueError(
                f"Invalid reasoning effort '{effort}' for agent '{agent_role}'. "
                f"Must be one of: {sorted(allowed_efforts)}."
            )

    # Log the successful validation
    logger.info("Successfully verified the reasoning effort schedule for all agents.")

    return True

# -------------------------------------------------------------------------------------------------------------------------------
# Task 8, Step 3: Freeze the max output tokens schedule per prompt
# -------------------------------------------------------------------------------------------------------------------------------

def validate_max_output_tokens_schedule(reasoning_effort_schedule: Dict[str, Any]) -> bool:
    """
    Validates that the max_output_tokens value for each agent is a strictly
    positive integer within safe operational bounds, preventing truncated
    outputs or runaway generation.

    Parameters
    ----------
    reasoning_effort_schedule : Dict[str, Any]
        The dictionary mapping agent roles to their token limit configurations.

    Returns
    -------
    bool
        True if all token limits are valid.

    Raises
    ------
    TypeError
        If the input is not a dictionary.
    ValueError
        If any agent role is missing a token limit, if the limit is not an
        integer, or if it falls outside the safe operational bounds.
    """
    # Validate that the input is a dictionary
    if not isinstance(reasoning_effort_schedule, dict):
        raise TypeError(f"reasoning_effort_schedule must be a dictionary, got {type(reasoning_effort_schedule).__name__}.")

    # Define the absolute maximum token ceiling to prevent runaway generation
    # 16384 is a generous but safe upper bound for structured financial outputs
    absolute_max_tokens: int = 16384

    # Iterate through the schedule to validate each agent's configuration
    for agent_role, config in reasoning_effort_schedule.items():
        # Ensure the configuration for the agent is a dictionary
        if not isinstance(config, dict):
            raise TypeError(f"Configuration for agent '{agent_role}' must be a dictionary.")

        # Extract the token limit
        max_tokens: Any = config.get("max_output_tokens")

        # Check if the token limit is missing
        if max_tokens is None:
            raise ValueError(f"Agent '{agent_role}' is missing the required 'max_output_tokens' key.")

        # Validate that the token limit is an integer and not a boolean
        if not isinstance(max_tokens, int) or isinstance(max_tokens, bool):
            raise ValueError(f"'max_output_tokens' for agent '{agent_role}' must be an integer, got {type(max_tokens).__name__}.")

        # Validate that the token limit is strictly positive
        if max_tokens <= 0:
            raise ValueError(f"'max_output_tokens' for agent '{agent_role}' must be strictly positive, got {max_tokens}.")

        # Validate that the token limit does not exceed the absolute ceiling
        if max_tokens > absolute_max_tokens:
            raise ValueError(
                f"'max_output_tokens' for agent '{agent_role}' ({max_tokens}) exceeds the "
                f"absolute safe ceiling of {absolute_max_tokens}."
            )

    # Log the successful validation
    logger.info("Successfully verified the max output tokens schedule for all agents.")

    return True

# -------------------------------------------------------------------------------------------------------------------------------
# Task 8, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_8_validate_gpt52_conventions(config: Dict[str, Any]) -> Dict[str, Any]:
    """
    Orchestrates the validation of GPT-5.2 prompting conventions, ensuring
    strict adherence to Responses API role discipline and compute allocation.
    Returns a deep-copied, validated configuration dictionary to maintain
    the functional data-passing paradigm.

    Parameters
    ----------
    config : Dict[str, Any]
        The study configuration dictionary.

    Returns
    -------
    Dict[str, Any]
        A deep copy of the validated configuration dictionary.

    Raises
    ------
    KeyError
        If required configuration keys are missing.
    """
    # Log the commencement of the GPT-5.2 conventions validation task
    logger.info("Starting Task 8: Validate GPT-5.2 prompting conventions (Responses API role discipline).")

    # Create a deep copy of the configuration to prevent unintended side effects on global state
    patched_config: Dict[str, Any] = copy.deepcopy(config)

    # Extract necessary dictionaries from the copied configuration
    try:
        llm_invocation_policy: Dict[str, Any] = patched_config["INFRASTRUCTURE"]["LLM_INVOCATION_POLICY"]
        reasoning_effort_schedule: Dict[str, Any] = patched_config["INFRASTRUCTURE"]["REASONING_EFFORT_SCHEDULE"]
    except KeyError as exc:
        # Raise a KeyError if the expected nested structures are missing
        raise KeyError(f"Configuration is missing required keys for Task 8: {exc}") from exc

    # Step 1: Freeze "system-level" instructions to the `developer` role
    validate_instruction_role(llm_invocation_policy)

    # Step 2: Freeze the reasoning effort schedule by agent type
    validate_reasoning_effort_schedule(reasoning_effort_schedule)

    # Step 3: Freeze the max output tokens schedule per prompt
    validate_max_output_tokens_schedule(reasoning_effort_schedule)

    # Log the successful completion of the orchestration task
    logger.info("Task 8 completed successfully. GPT-5.2 prompting conventions are strictly enforced.")

    # Return the deep-copied, validated configuration dictionary to maintain functional chaining
    return patched_config
````

````python
# Task 9 — Validate schemas and indices for all ingested dataframes

# ==============================================================================
# Task 9: Validate schemas and indices for all ingested dataframes
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 9, Step 1: Validate index type, required columns, dtypes, and monotonic increasing time
# -------------------------------------------------------------------------------------------------------------------------------

def validate_dataframe_schema(
    df: pd.DataFrame,
    schema_config: Dict[str, Any],
    df_name: str
) -> bool:
    """
    Validates the structural integrity of an ingested dataframe, ensuring its index type,
    column presence, data types, and chronological ordering strictly adhere to the
    defined configuration schema.

    Parameters
    ----------
    df : pd.DataFrame
        The raw ingested pandas DataFrame to be validated.
    schema_config : Dict[str, Any]
        The schema configuration dictionary for this specific dataframe, containing
        'index' and 'schema' definitions.
    df_name : str
        The string identifier of the dataframe for logging and error reporting.

    Returns
    -------
    bool
        True if the dataframe passes all structural and type validations.

    Raises
    ------
    TypeError
        If the input is not a DataFrame, or if the index/column types do not match the schema.
    ValueError
        If required columns are missing, or if the temporal index is not monotonically increasing.
    """
    # Verify that the input object is indeed a pandas DataFrame
    if not isinstance(df, pd.DataFrame):
        # Raise a TypeError if the input is of an incorrect type
        raise TypeError(f"Expected {df_name} to be a pandas DataFrame, got {type(df).__name__}.")

    # Extract the expected index definition from the schema configuration
    expected_index_def: Any = schema_config.get("index")

    # Extract the expected column schema dictionary from the configuration
    expected_schema: Dict[str, str] = schema_config.get("schema", {})

    # Validate MultiIndex structures if the expected index definition is a list
    if isinstance(expected_index_def, list):
        # Verify that the dataframe actually possesses a MultiIndex
        if not isinstance(df.index, pd.MultiIndex):
            # Raise a TypeError if a MultiIndex was expected but not found
            raise TypeError(f"{df_name} must have a MultiIndex with levels {expected_index_def}.")

        # Extract the actual names of the MultiIndex levels
        actual_index_names: List[str] = list(df.index.names)

        # Verify that the actual index names exactly match the expected index names
        if actual_index_names != expected_index_def:
            # Raise a ValueError if the MultiIndex levels are incorrectly named or ordered
            raise ValueError(f"{df_name} MultiIndex levels must be {expected_index_def}, got {actual_index_names}.")

        # Extract the temporal level (assumed to be 'date') for chronological validation
        if "date" in actual_index_names:
            # Extract the date level values
            date_level: pd.Index = df.index.get_level_values("date")
            # Verify that the date level is a DatetimeIndex
            if not isinstance(date_level, pd.DatetimeIndex):
                # Raise a TypeError if the date level is not properly typed as datetime
                raise TypeError(f"The 'date' level in {df_name} MultiIndex must be a DatetimeIndex.")
            # Verify that the dates are monotonically increasing to prevent chronological corruption
            if not date_level.is_monotonic_increasing:
                # Raise a ValueError if the time series flows backward or scrambles
                raise ValueError(f"The 'date' level in {df_name} must be monotonically increasing.")

    # Validate standard DatetimeIndex structures if the expected index is a string definition
    elif isinstance(expected_index_def, str) and "DatetimeIndex" in expected_index_def:
        # Verify that the dataframe possesses a standard DatetimeIndex
        if not isinstance(df.index, pd.DatetimeIndex):
            # Raise a TypeError if a DatetimeIndex was expected but not found
            raise TypeError(f"{df_name} must have a DatetimeIndex.")
        # Verify that the dates are monotonically increasing
        if not df.index.is_monotonic_increasing:
            # Raise a ValueError if the time series is not chronologically sorted
            raise ValueError(f"The DatetimeIndex in {df_name} must be monotonically increasing.")
    else:
        # Raise a ValueError if the schema configuration provides an unrecognized index definition
        raise ValueError(f"Unrecognized index definition in schema for {df_name}: {expected_index_def}")

    # Extract the actual columns present in the dataframe
    actual_columns: set = set(df.columns)

    # Extract the expected columns from the schema dictionary
    expected_columns: set = set(expected_schema.keys())

    # Identify any required columns that are missing from the dataframe
    missing_columns: set = expected_columns - actual_columns

    # Check if there are any missing columns
    if missing_columns:
        # Raise a ValueError detailing the exact missing columns
        raise ValueError(f"{df_name} is missing required columns: {sorted(missing_columns)}")

    # Iterate through the expected schema to validate the data type of each column
    for col_name, expected_dtype in expected_schema.items():
        # Extract the actual pandas Series for the column
        col_series: pd.Series = df[col_name]

        # Validate float64 expectations
        if expected_dtype == "float64":
            # Use pandas API to strictly verify the column is a float type
            if not is_float_dtype(col_series):
                # Raise a TypeError if the column was silently cast to object or int
                raise TypeError(f"Column '{col_name}' in {df_name} must be float64, got {col_series.dtype}.")

        # Validate int64 expectations
        elif expected_dtype == "int64":
            # Use pandas API to strictly verify the column is an integer type
            if not is_integer_dtype(col_series):
                # Raise a TypeError if the column is not an integer
                raise TypeError(f"Column '{col_name}' in {df_name} must be int64, got {col_series.dtype}.")

        # Validate object/string expectations
        elif expected_dtype == "object":
            # Use pandas API to strictly verify the column is an object type
            if not is_object_dtype(col_series):
                # Raise a TypeError if the column is not an object
                raise TypeError(f"Column '{col_name}' in {df_name} must be object/string, got {col_series.dtype}.")

        else:
            # Raise a ValueError if the schema requests an unsupported data type validation
            raise ValueError(f"Unsupported dtype '{expected_dtype}' specified in schema for column '{col_name}'.")

    # Log the successful structural and type validation of the dataframe
    logger.info(f"Successfully validated schema, index, and dtypes for {df_name}.")

    # Return True to indicate all validations passed
    return True

# -------------------------------------------------------------------------------------------------------------------------------
# Task 9, Step 2: Validate uniqueness constraints
# -------------------------------------------------------------------------------------------------------------------------------

def validate_dataframe_uniqueness(df: pd.DataFrame, df_name: str) -> bool:
    """
    Validates that the dataframe index contains no duplicate entries, ensuring
    relational integrity and preventing explosive joins in downstream operations.

    Parameters
    ----------
    df : pd.DataFrame
        The pandas DataFrame whose index uniqueness is to be validated.
    df_name : str
        The string identifier of the dataframe for logging and error reporting.

    Returns
    -------
    bool
        True if the dataframe index is strictly unique.

    Raises
    ------
    ValueError
        If duplicate index entries are detected.
    """
    # Evaluate the duplicated method on the dataframe index to find any repeated entries
    duplicates_mask: np.ndarray = df.index.duplicated()

    # Check if any duplicates exist in the boolean mask
    if duplicates_mask.any():
        # Extract the specific index values that are duplicated for error reporting
        duplicate_entries: pd.Index = df.index[duplicates_mask].unique()

        # Raise a fatal ValueError detailing the exact duplicate entries
        raise ValueError(
            f"Fatal Data Integrity Error: {df_name} contains duplicate index entries. "
            f"Uniqueness constraint violated. Duplicates found for: {duplicate_entries.tolist()}"
        )

    # Log the successful validation of index uniqueness
    logger.info(f"Successfully validated index uniqueness constraint for {df_name}.")

    # Return True to indicate the uniqueness constraint is satisfied
    return True

# -------------------------------------------------------------------------------------------------------------------------------
# Task 9, Step 3: Validate "as-of" coverage and enforce point-in-time discipline
# -------------------------------------------------------------------------------------------------------------------------------

def enforce_point_in_time(df: pd.DataFrame, as_of_date_str: str, df_name: str) -> pd.DataFrame:
    """
    Enforces point-in-time discipline by verifying sufficient data coverage up to
    the as-of date, and strictly truncating any observations that leak from the future.
    Dynamically handles timezone-aware indices to prevent comparison crashes.

    Parameters
    ----------
    df : pd.DataFrame
        The pandas DataFrame to be truncated.
    as_of_date_str : str
        The ISO-8601 string representation of the as-of date.
    df_name : str
        The string identifier of the dataframe for logging and error reporting.

    Returns
    -------
    pd.DataFrame
        The truncated pandas DataFrame containing no future data.

    Raises
    ------
    ValueError
        If the dataframe does not contain data up to the required as-of date.
    """
    # Parse the ISO-8601 string into a pandas Timestamp object for comparison
    as_of_ts: pd.Timestamp = pd.Timestamp(as_of_date_str)

    # Determine the temporal index for comparison based on the index structure
    if isinstance(df.index, pd.MultiIndex):
        # Extract the 'date' level from the MultiIndex
        temporal_index: pd.DatetimeIndex = df.index.get_level_values("date")
    else:
        # Use the standard DatetimeIndex directly
        temporal_index = df.index

    # Dynamically inspect the timezone of the dataframe's temporal index
    if temporal_index.tz is not None:
        # If the dataframe is timezone-aware, localize the as-of timestamp to match that exact timezone
        # This prevents the fatal TypeError: Cannot compare tz-naive and tz-aware timestamps
        as_of_ts = as_of_ts.tz_localize(temporal_index.tz)

    # Extract the maximum (most recent) date available in the dataset
    max_available_date: pd.Timestamp = temporal_index.max()

    # Verify that the dataset contains information up to or beyond the as-of date
    if max_available_date < as_of_ts:
        # Raise a ValueError if the dataset is stale and cannot support the simulation
        raise ValueError(
            f"Insufficient data coverage in {df_name}. "
            f"Maximum available date is {max_available_date.date()}, but AS_OF_DATE is {as_of_ts.date()}."
        )

    # Create a boolean mask to identify rows that occur on or before the as-of date
    past_and_present_mask: np.ndarray = temporal_index <= as_of_ts

    # Apply the boolean mask to hard-slice the dataframe, dropping all future rows
    truncated_df: pd.DataFrame = df.loc[past_and_present_mask].copy()

    # Calculate the number of future rows that were dropped to prevent lookahead bias
    rows_dropped: int = len(df) - len(truncated_df)

    # Log the enforcement of point-in-time discipline
    if rows_dropped > 0:
        logger.info(f"Enforced point-in-time discipline on {df_name}: Dropped {rows_dropped} future observations.")
    else:
        logger.info(f"Point-in-time check passed for {df_name}: No future observations detected.")

    # Return the strictly truncated dataframe
    return truncated_df

# -------------------------------------------------------------------------------------------------------------------------------
# Task 9, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_9_validate_ingested_dataframes(
    raw_dataframes: Dict[str, pd.DataFrame],
    config: Dict[str, Any]
) -> Dict[str, pd.DataFrame]:
    """
    Orchestrates the comprehensive validation of all ingested dataframes, ensuring
    schema compliance, index uniqueness, and strict point-in-time truncation.

    Parameters
    ----------
    raw_dataframes : Dict[str, pd.DataFrame]
        A dictionary mapping dataframe string identifiers to their raw pandas DataFrame objects.
    config : Dict[str, Any]
        The validated study configuration dictionary containing RAW_DATA_SCHEMAS and AS_OF_DATE.

    Returns
    -------
    Dict[str, pd.DataFrame]
        A dictionary mapping dataframe string identifiers to their validated and truncated DataFrame objects.

    Raises
    ------
    KeyError
        If required dataframes are missing from the input dictionary or if configuration keys are missing.
    """
    # Log the commencement of the dataframe validation task
    logger.info("Starting Task 9: Validate schemas and indices for all ingested dataframes.")

    # Extract the raw data schemas registry from the configuration
    try:
        schemas_registry: Dict[str, Any] = config["RAW_DATA_SCHEMAS"]
        as_of_date_str: str = config["AS_OF_DATE"]
    except KeyError as exc:
        # Raise a KeyError if the expected configuration structures are missing
        raise KeyError(f"Configuration is missing required keys for Task 9: {exc}") from exc

    # Initialize an empty dictionary to store the validated and truncated dataframes
    validated_dataframes: Dict[str, pd.DataFrame] = {}

    # Iterate through the expected schemas defined in the configuration
    for schema_key, schema_config in schemas_registry.items():
        # Extract the expected dataframe name from the schema configuration
        df_name: str = schema_config.get("dataframe_name", schema_key)

        # Verify that the expected dataframe was provided in the input dictionary
        if df_name not in raw_dataframes:
            # Raise a KeyError if a required dataset is missing from the ingestion payload
            raise KeyError(f"Required dataframe '{df_name}' was not provided in the raw_dataframes input.")

        # Extract the raw dataframe object
        raw_df: pd.DataFrame = raw_dataframes[df_name]

        # Step 1: Validate index type, required columns, dtypes, and monotonic increasing time
        validate_dataframe_schema(raw_df, schema_config, df_name)

        # Step 2: Validate uniqueness constraints to prevent relational corruption
        validate_dataframe_uniqueness(raw_df, df_name)

        # Step 3: Validate as-of coverage and enforce point-in-time discipline by truncating future data
        truncated_df: pd.DataFrame = enforce_point_in_time(raw_df, as_of_date_str, df_name)

        # Store the validated and truncated dataframe in the output dictionary
        validated_dataframes[df_name] = truncated_df

    # Log the successful completion of the orchestration task
    logger.info("Task 9 completed successfully. All ingested dataframes are validated and point-in-time compliant.")

    # Return the dictionary of validated dataframes
    return validated_dataframes
````

````python
# Task 10 — Cleanse time indices and align calendars (foundational cleansing)

# ==============================================================================
# Task 10: Cleanse time indices and align calendars (foundational cleansing)
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 10, Step 1: Normalize timestamps (timezone-naive UTC-equivalent and month-end labeling)
# -------------------------------------------------------------------------------------------------------------------------------

def normalize_timestamps(
    df: pd.DataFrame,
    enforce_month_end: bool = False,
    date_level_name: str = "date"
) -> pd.DataFrame:
    """
    Normalizes the temporal index of a dataframe by stripping timezone information
    to ensure a canonical UTC-naive representation, and snapping all times to exactly
    midnight (00:00:00). Optionally snaps dates to the exact month-end.

    This function must be executed BEFORE point-in-time truncation to prevent
    timezone-aware comparison crashes.

    Parameters
    ----------
    df : pd.DataFrame
        The input dataframe with a DatetimeIndex or a MultiIndex containing a date level.
    enforce_month_end : bool, optional
        If True, applies a MonthEnd offset to snap all dates to the end of the month.
        Defaults to False.
    date_level_name : str, optional
        The name of the temporal level if the index is a MultiIndex. Defaults to "date".

    Returns
    -------
    pd.DataFrame
        A new dataframe with normalized, midnight-aligned timestamps.

    Raises
    ------
    TypeError
        If the input is not a pandas DataFrame or lacks a valid temporal index.
    """
    # Create a deep copy of the dataframe to preserve immutability of the input
    df_clean: pd.DataFrame = df.copy(deep=True)

    # Extract the index from the copied dataframe
    idx: pd.Index = df_clean.index

    # Check if the index is a MultiIndex structure
    if isinstance(idx, pd.MultiIndex):
        # Verify that the specified date level exists in the MultiIndex
        if date_level_name not in idx.names:
            # Raise a TypeError if the temporal level cannot be found
            raise TypeError(f"MultiIndex must contain a level named '{date_level_name}'.")

        # Extract the temporal level values from the MultiIndex
        date_level: pd.DatetimeIndex = idx.get_level_values(date_level_name)

        # Check if the temporal level has timezone information attached
        if date_level.tz is not None:
            # Strip the timezone information using tz_convert(None) for tz-aware indices
            date_level = date_level.tz_convert(None)

        # Normalize the timestamps to exactly midnight (00:00:00) to prevent intra-day merge failures
        date_level = date_level.normalize()

        # Check if month-end snapping is requested
        if enforce_month_end:
            # Apply the MonthEnd offset to snap all dates to the end of their respective months
            date_level = date_level + MonthEnd(0)

        # Reassign the normalized temporal level back to the MultiIndex
        df_clean.index = idx.set_levels(date_level, level=date_level_name)

    # Handle the case where the index is a standard DatetimeIndex
    elif isinstance(idx, pd.DatetimeIndex):
        # Check if the DatetimeIndex has timezone information attached
        if idx.tz is not None:
            # Strip the timezone information using tz_convert(None)
            df_clean.index = idx.tz_convert(None)

        # Normalize the timestamps to exactly midnight (00:00:00)
        df_clean.index = df_clean.index.normalize()

        # Check if month-end snapping is requested
        if enforce_month_end:
            # Apply the MonthEnd offset to snap all dates to the end of their respective months
            df_clean.index = df_clean.index + MonthEnd(0)

    # Handle the case where the index is neither a MultiIndex nor a DatetimeIndex
    else:
        # Raise a TypeError because temporal normalization requires a temporal index
        raise TypeError(f"Dataframe index must be a DatetimeIndex or MultiIndex, got {type(idx).__name__}.")

    # Log the successful normalization of the timestamps
    logger.debug(f"Normalized timestamps to midnight. Month-end enforced: {enforce_month_end}.")

    # Return the dataframe with the cleansed and normalized index
    return df_clean

# -------------------------------------------------------------------------------------------------------------------------------
# Task 10, Step 2: Align daily data to monthly SAA calendar
# -------------------------------------------------------------------------------------------------------------------------------

def align_daily_to_monthly(
    df_daily: pd.DataFrame,
    returns_cols: List[str],
    rates_cols: List[str],
    resample_rule: str = "BME"
) -> pd.DataFrame:
    """
    Aggregates daily factor returns and rates into a monthly frequency.
    Applies geometric compounding to returns and point-in-time sampling to rates.

    CRITICAL REMEDIATION: Injects `min_count=1` into the `.prod()` aggregation.
    This mathematically enforces that the product of an empty set (an all-NaN month)
    yields NaN rather than the default 1.0, preventing the hallucination of 0.0% returns.

    Equation for returns: R_{monthly} = \prod_{d=1}^{D} (1 + r_{daily, d}) - 1

    Parameters
    ----------
    df_daily : pd.DataFrame
        The daily frequency dataframe containing factors and rates.
    returns_cols : List[str]
        List of column names representing daily returns requiring geometric compounding.
    rates_cols : List[str]
        List of column names representing daily rates requiring point-in-time sampling.
    resample_rule : str, optional
        The pandas frequency alias for resampling. Defaults to "BME" (Business Month End).

    Returns
    -------
    pd.DataFrame
        A new dataframe aggregated to the specified monthly frequency.

    Raises
    ------
    TypeError
        If the input is not a DataFrame or lacks a DatetimeIndex.
    ValueError
        If specified columns are missing from the dataframe.
    """
    # Verify that the input is a pandas DataFrame
    if not isinstance(df_daily, pd.DataFrame):
        raise TypeError(f"df_daily must be a pandas DataFrame, got {type(df_daily).__name__}.")

    # Verify that the dataframe possesses a standard DatetimeIndex
    if not isinstance(df_daily.index, pd.DatetimeIndex):
        raise TypeError("df_daily must have a DatetimeIndex for resampling.")

    # Combine the lists of required columns
    all_required_cols: List[str] = returns_cols + rates_cols

    # Identify any required columns that are missing from the dataframe
    missing_cols: List[str] = [col for col in all_required_cols if col not in df_daily.columns]

    if missing_cols:
        raise ValueError(f"df_daily is missing required columns for alignment: {missing_cols}")

    # Initialize an empty list to hold the aggregated series
    aggregated_series: List[pd.Series] = []

    # Process the columns that represent returns requiring geometric compounding
    if returns_cols:
        # Extract the subset of the dataframe containing only the return columns
        df_returns: pd.DataFrame = df_daily[returns_cols].copy()

        # Add 1.0 to all daily returns to prepare for geometric compounding
        df_gross: pd.DataFrame = df_returns + 1.0

        # Resample to the target frequency and compute the product of the gross returns
        # min_count=1 ensures that if a month has 0 valid observations, it returns NaN, not 1.0
        df_monthly_gross: pd.DataFrame = df_gross.resample(resample_rule).prod(min_count=1)

        # Subtract 1.0 to convert the compounded gross returns back to net periodic returns
        df_monthly_net: pd.DataFrame = df_monthly_gross - 1.0

        # Iterate over the processed return columns and append to the collection list
        for col in returns_cols:
            aggregated_series.append(df_monthly_net[col])

    # Process the columns that represent rates requiring point-in-time sampling
    if rates_cols:
        # Extract the subset of the dataframe containing only the rate columns
        df_rates: pd.DataFrame = df_daily[rates_cols].copy()

        # Resample to the target frequency and extract the last valid observation
        df_monthly_rates: pd.DataFrame = df_rates.resample(resample_rule).last()

        # Iterate over the processed rate columns and append to the collection list
        for col in rates_cols:
            aggregated_series.append(df_monthly_rates[col])

    # Concatenate all aggregated series into a single wide dataframe
    df_aligned: pd.DataFrame = pd.concat(aggregated_series, axis=1)

    # Log the successful alignment of the daily data to a monthly calendar
    logger.info(f"Successfully aligned daily data to monthly using rule '{resample_rule}' with strict NaN propagation.")

    # Return the temporally aligned dataframe
    return df_aligned

# -------------------------------------------------------------------------------------------------------------------------------
# Task 10, Step 3: Emit a "calendar audit" report
# -------------------------------------------------------------------------------------------------------------------------------

def generate_calendar_audit_report(
    dfs_metadata: List[Dict[str, Any]],
    output_dir: Path
) -> Dict[str, Any]:
    """
    Generates a calendar audit report by comparing the actual indices of dataframes
    against an expected continuous temporal grid, identifying any missing periods.

    Parameters
    ----------
    dfs_metadata : List[Dict[str, Any]]
        A list of dictionaries, each containing 'name' (str), 'df' (pd.DataFrame),
        'freq' (str), and 'date_level' (str, optional).
    output_dir : Path
        The directory where the audit report JSON will be saved.

    Returns
    -------
    Dict[str, Any]
        A dictionary containing the audit results for all evaluated dataframes.

    Raises
    ------
    OSError
        If the audit report cannot be written to the specified directory.
    """
    # Initialize the master dictionary to hold the audit results
    audit_report: Dict[str, Any] = {}

    # Iterate over the metadata for each dataframe to be audited
    for meta in dfs_metadata:
        # Extract the dataframe identifier
        df_name: str = meta["name"]

        # Extract the dataframe object
        df: pd.DataFrame = meta["df"]

        # Extract the expected frequency alias
        freq: str = meta["freq"]

        # Extract the date level name, defaulting to "date"
        date_level: str = meta.get("date_level", "date")

        # Extract the temporal index based on the index structure
        if isinstance(df.index, pd.MultiIndex):
            # Extract the unique dates from the specified level of the MultiIndex
            actual_dates: pd.DatetimeIndex = pd.DatetimeIndex(df.index.get_level_values(date_level).unique())
        else:
            # Use the standard DatetimeIndex directly
            actual_dates = pd.DatetimeIndex(df.index.unique())

        # Check if the extracted index is empty
        if actual_dates.empty:
            # Record that the dataframe contains no temporal data
            audit_report[df_name] = {"status": "empty", "missing_dates": []}
            # Skip further processing for this dataframe
            continue

        # Identify the earliest date in the actual index
        min_date: pd.Timestamp = actual_dates.min()

        # Identify the latest date in the actual index
        max_date: pd.Timestamp = actual_dates.max()

        # Generate the expected continuous calendar grid between the min and max dates
        expected_dates: pd.DatetimeIndex = pd.date_range(start=min_date, end=max_date, freq=freq)

        # Compute the set difference to find dates present in the expected grid but missing in reality
        missing_dates_idx: pd.DatetimeIndex = expected_dates.difference(actual_dates)

        # Convert the missing pandas Timestamps into ISO-8601 formatted strings
        missing_dates_str: List[str] = [ts.strftime("%Y-%m-%d") for ts in missing_dates_idx]

        # Populate the audit report for this specific dataframe
        audit_report[df_name] = {
            "min_date": min_date.strftime("%Y-%m-%d"),
            "max_date": max_date.strftime("%Y-%m-%d"),
            "expected_frequency": freq,
            "missing_count": len(missing_dates_str),
            "missing_dates": missing_dates_str
        }

    # Ensure the output directory exists, creating it if necessary
    output_dir.mkdir(parents=True, exist_ok=True)

    # Construct the full file path for the audit report JSON
    report_path: Path = output_dir / "calendar_audit_report.json"

    # Attempt to write the audit report to disk
    try:
        # Open the file in write mode with UTF-8 encoding
        with open(report_path, "w", encoding="utf-8") as f:
            # Serialize the dictionary to a formatted JSON string and write it
            json.dump(audit_report, f, indent=4)
    except OSError as exc:
        # Raise an OSError if filesystem permissions prevent writing
        raise OSError(f"Failed to write calendar audit report to {report_path}: {exc}") from exc

    # Log the successful generation and persistence of the audit report
    logger.info(f"Calendar audit report generated and saved to {report_path}.")

    # Return the audit report dictionary
    return audit_report

# -------------------------------------------------------------------------------------------------------------------------------
# Task 10, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_10_cleanse_time_indices(
    validated_dataframes: Dict[str, pd.DataFrame],
    config: Dict[str, Any],
    run_output_dir: Path
) -> Dict[str, pd.DataFrame]:
    """
    Orchestrates the cleansing of time indices and alignment of calendars across
    all ingested dataframes, ensuring a canonical temporal grid.

    Parameters
    ----------
    validated_dataframes : Dict[str, pd.DataFrame]
        The dictionary of dataframes that passed Task 9 validation.
    config : Dict[str, Any]
        The study configuration dictionary containing frequency conventions.
    run_output_dir : Path
        The root directory for the current run's artifacts.

    Returns
    -------
    Dict[str, pd.DataFrame]
        A dictionary of temporally cleansed and aligned dataframes.

    Raises
    ------
    KeyError
        If required configuration keys or dataframes are missing.
    """
    # Log the commencement of the temporal cleansing task
    logger.info("Starting Task 10: Cleanse time indices and align calendars.")

    # Extract the frequency conventions from the configuration
    try:
        freq_config: Dict[str, Any] = config["DATA_CONVENTIONS"]["frequency"]
        resample_rules: Dict[str, Any] = config["DATA_CONVENTIONS"]["resampling_rules"]
    except KeyError as exc:
        # Raise a KeyError if the expected nested structures are missing
        raise KeyError(f"Configuration is missing required keys for Task 10: {exc}") from exc

    # Initialize an empty dictionary to store the cleansed dataframes
    cleansed_dataframes: Dict[str, pd.DataFrame] = {}

    # Initialize a list to collect metadata for the calendar audit report
    audit_metadata: List[Dict[str, Any]] = []

    # Iterate over the validated dataframes
    for df_name, df in validated_dataframes.items():
        # Determine if the dataframe represents monthly data based on its name
        # This is a heuristic mapping based on the architecture's data schemas
        is_monthly: bool = df_name in [
            "df_macro_raw", "df_total_return_raw", "df_fundamentals_raw",
            "df_signals_raw", "df_survey_cma_raw", "df_fixed_income_curves_spreads_raw",
            "df_cpi_index_raw"
        ]

        # Step 1: Normalize timestamps, enforcing month-end snapping for monthly data
        df_norm: pd.DataFrame = normalize_timestamps(df, enforce_month_end=is_monthly)

        # Determine the expected frequency alias for the audit report
        expected_freq: str = "ME" if is_monthly else "B"

        # Step 2: Align daily factor data to the monthly SAA calendar
        if df_name == "df_benchmark_factors_raw":
            # Extract the resampling rule from the configuration
            rule: str = resample_rules.get("daily_to_monthly", "BME")

            # Define the columns requiring geometric compounding
            returns_cols: List[str] = ["mkt_rf", "smb", "hml"]

            # Define the columns requiring point-in-time sampling
            rates_cols: List[str] = ["rf"]

            # Execute the alignment transformation
            df_aligned: pd.DataFrame = align_daily_to_monthly(df_norm, returns_cols, rates_cols, resample_rule=rule)

            # Store the aligned dataframe
            cleansed_dataframes[df_name] = df_aligned

            # Update the expected frequency for the audit report since it is now monthly
            expected_freq = rule

            # Append metadata for the audit report
            audit_metadata.append({"name": df_name, "df": df_aligned, "freq": expected_freq})

        else:
            # Store the normalized dataframe directly if no daily-to-monthly alignment is needed
            cleansed_dataframes[df_name] = df_norm

            # Append metadata for the audit report
            audit_metadata.append({"name": df_name, "df": df_norm, "freq": expected_freq})

    # Step 3: Emit the calendar audit report to verify temporal continuity
    generate_calendar_audit_report(audit_metadata, run_output_dir)

    # Log the successful completion of the orchestration task
    logger.info("Task 10 completed successfully. All time indices are cleansed and aligned.")

    # Return the dictionary of temporally cleansed dataframes
    return cleansed_dataframes
````

````python
# Task 11 — Cleanse df_total_return_raw and construct monthly returns panel

# ==============================================================================
# Task 11: Cleanse df_total_return_raw and construct monthly returns panel
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 11, Step 1: Validate that total_return_index is strictly positive
# -------------------------------------------------------------------------------------------------------------------------------

def validate_total_return_strictly_positive(df_total_return: pd.DataFrame) -> pd.DataFrame:
    """
    Validates that the 'total_return_index' column contains strictly positive values,
    enforcing financial data invariants and preventing downstream mathematical errors.

    Parameters
    ----------
    df_total_return : pd.DataFrame
        The temporally cleansed total return dataframe containing the 'total_return_index' column.

    Returns
    -------
    pd.DataFrame
        The validated dataframe, guaranteed to contain only strictly positive index values.

    Raises
    ------
    TypeError
        If the input is not a pandas DataFrame.
    ValueError
        If the 'total_return_index' column is missing or contains values <= 0.
    """
    # Verify that the input is a pandas DataFrame
    if not isinstance(df_total_return, pd.DataFrame):
        # Raise a TypeError if the input is invalid
        raise TypeError(f"df_total_return must be a pandas DataFrame, got {type(df_total_return).__name__}.")

    # Verify that the required column exists in the dataframe
    if "total_return_index" not in df_total_return.columns:
        # Raise a ValueError if the target column is missing
        raise ValueError("df_total_return must contain the column 'total_return_index'.")

    # Extract the total return index series for evaluation
    tri_series: pd.Series = df_total_return["total_return_index"]

    # Evaluate the series to find any values that are less than or equal to zero
    non_positive_mask: pd.Series = tri_series <= 0

    # Check if any non-positive values exist
    if non_positive_mask.any():
        # Extract the specific index locations of the invalid data for error reporting
        invalid_entries: pd.Index = df_total_return[non_positive_mask].index
        # Raise a fatal ValueError detailing the exact invalid entries
        raise ValueError(
            f"Fatal Data Integrity Error: 'total_return_index' contains non-positive values. "
            f"Total return indices must be strictly positive. Invalid entries found at: {invalid_entries.tolist()}"
        )

    # Log the successful validation of the strictly positive invariant
    logger.info("Successfully validated that all 'total_return_index' values are strictly positive.")

    # Return the validated dataframe
    return df_total_return

# -------------------------------------------------------------------------------------------------------------------------------
# Task 11, Step 2: Compute monthly returns per ticker
# -------------------------------------------------------------------------------------------------------------------------------

def compute_monthly_returns_per_ticker(df_total_return_valid: pd.DataFrame) -> pd.Series:
    """
    Computes discrete simple periodic returns for each asset in the panel,
    strictly preventing cross-sectional leakage and disabling default forward-filling.

    Equation: r_{t,i} = (TR_{t,i} / TR_{t-1,i}) - 1

    Parameters
    ----------
    df_total_return_valid : pd.DataFrame
        The validated total return dataframe with strictly positive index values.

    Returns
    -------
    pd.Series
        A pandas Series containing the computed discrete periodic returns, retaining
        the original MultiIndex structure.

    Raises
    ------
    TypeError
        If the input dataframe does not possess a MultiIndex.
    ValueError
        If the required MultiIndex levels ('ticker', 'investment_universe', 'date') are missing.
    """
    # Verify that the dataframe possesses a MultiIndex
    if not isinstance(df_total_return_valid.index, pd.MultiIndex):
        # Raise a TypeError if the index structure is invalid for panel data
        raise TypeError("df_total_return_valid must have a MultiIndex.")

    # Define the required levels for cross-sectional and temporal grouping
    required_levels: List[str] = ["date", "ticker", "investment_universe"]

    # Iterate through the required levels to ensure they exist in the MultiIndex
    for level in required_levels:
        if level not in df_total_return_valid.index.names:
            # Raise a ValueError if a critical grouping level is missing
            raise ValueError(f"MultiIndex must contain the level '{level}'.")

    # Create a deep copy to preserve the immutability of the input dataframe
    df_calc: pd.DataFrame = df_total_return_valid.copy(deep=True)

    # Sort the MultiIndex first by cross-sectional identifiers, then chronologically
    # This guarantees that the shift operation computes returns within the same asset over time
    df_calc = df_calc.sort_index(level=["ticker", "investment_universe", "date"])

    # Group the dataframe by the cross-sectional identifiers
    grouped_data = df_calc.groupby(level=["ticker", "investment_universe"])

    # Compute the percentage change chronologically, explicitly disabling forward-fill
    # fill_method=None ensures that missing data results in NaN returns, preserving true volatility
    periodic_returns: pd.Series = grouped_data["total_return_index"].pct_change(fill_method=None)

    # Drop the initial NaN values generated by the percentage change shift operation
    periodic_returns = periodic_returns.dropna()

    # Name the resulting series for clarity in downstream operations
    periodic_returns.name = "periodic_return"

    # Log the successful computation of the periodic returns
    logger.info("Successfully computed discrete periodic returns per ticker without forward-filling.")

    # Return the computed returns series
    return periodic_returns

# -------------------------------------------------------------------------------------------------------------------------------
# Task 11, Step 3: Create a wide returns matrix aligned to the 18-asset universe
# -------------------------------------------------------------------------------------------------------------------------------

def construct_wide_returns_matrix(
    periodic_returns: pd.Series,
    universe_map: Dict[str, Dict[str, Any]],
    canonical_order: List[str]
) -> pd.DataFrame:
    """
    Transforms the long-format returns series into a wide T x N matrix, maps tickers
    to canonical IPS asset class names, and strictly enforces the 18-asset dimensionality.

    Parameters
    ----------
    periodic_returns : pd.Series
        The long-format series of computed periodic returns with a MultiIndex.
    universe_map : Dict[str, Dict[str, Any]]
        The mapping dictionary linking canonical asset class names to their tickers.
    canonical_order : List[str]
        The strict, ordered list of the 18 canonical asset class names defined in the IPS.

    Returns
    -------
    pd.DataFrame
        A wide-format pandas DataFrame of shape (T, 18) representing the returns matrix.

    Raises
    ------
    ValueError
        If the resulting matrix does not have exactly 18 columns.
    """
    # Create a reverse mapping dictionary from ticker to canonical IPS asset class name
    ticker_to_ips: Dict[str, str] = {
        meta.get("ticker"): asset_class
        for asset_class, meta in universe_map.items()
        if meta.get("ticker")
    }

    # Reset the 'investment_universe' level from the MultiIndex as it is no longer needed for reshaping
    # This leaves a two-level MultiIndex of ['date', 'ticker']
    returns_reset: pd.Series = periodic_returns.reset_index(level="investment_universe", drop=True)

    # Unstack the 'ticker' level to pivot the data from long format to wide format (T x N)
    wide_returns: pd.DataFrame = returns_reset.unstack(level="ticker")

    # Rename the columns from raw tickers to canonical IPS asset class names using the reverse mapping
    wide_returns = wide_returns.rename(columns=ticker_to_ips)

    # Reindex the columns to strictly match the canonical 18-asset order
    # This forces the creation of exactly 18 columns; missing assets will be filled with NaNs
    aligned_returns_matrix: pd.DataFrame = wide_returns.reindex(columns=canonical_order)

    # Extract the number of columns in the resulting matrix
    num_columns: int = aligned_returns_matrix.shape[1]

    # Assert that the matrix dimensionality strictly matches the 18-asset universe requirement
    if num_columns != 18:
        # Raise a fatal ValueError if the dimensionality constraint is violated
        raise ValueError(f"Fatal Dimensionality Error: Returns matrix must have exactly 18 columns, got {num_columns}.")

    # Log the successful construction and alignment of the wide returns matrix
    logger.info(f"Successfully constructed wide returns matrix with shape {aligned_returns_matrix.shape}.")

    # Return the dimensionally strict returns matrix
    return aligned_returns_matrix

# -------------------------------------------------------------------------------------------------------------------------------
# Task 11, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_11_construct_monthly_returns_panel(
    df_total_return_raw: pd.DataFrame,
    universe_map: Dict[str, Dict[str, Any]],
    config: Dict[str, Any]
) -> pd.DataFrame:
    """
    Orchestrates the cleansing of the total return data and the construction of the
    canonical T x 18 monthly returns matrix.

    Parameters
    ----------
    df_total_return_raw : pd.DataFrame
        The temporally cleansed raw total return dataframe.
    universe_map : Dict[str, Dict[str, Any]]
        The mapping dictionary linking canonical asset class names to their tickers.
    config : Dict[str, Any]
        The validated study configuration dictionary containing the IPS asset classes.

    Returns
    -------
    pd.DataFrame
        The canonical wide-format monthly returns matrix of shape (T, 18).

    Raises
    ------
    KeyError
        If required configuration keys are missing.
    """
    # Log the commencement of the returns panel construction task
    logger.info("Starting Task 11: Cleanse df_total_return_raw and construct monthly returns panel.")

    # Extract the canonical asset class order from the configuration
    try:
        asset_classes_dict: Dict[str, List[str]] = config["IPS_GOVERNANCE"]["ASSET_CLASSES"]
        # Flatten the dictionary into a canonical ordered list
        canonical_order: List[str] = []
        for category in ["equity", "fixed_income", "real_assets", "cash"]:
            canonical_order.extend(asset_classes_dict.get(category, []))
    except KeyError as exc:
        # Raise a KeyError if the expected nested structures are missing
        raise KeyError(f"Configuration is missing required IPS_GOVERNANCE sub-keys: {exc}") from exc

    # Step 1: Validate that the total return index is strictly positive
    df_total_return_valid: pd.DataFrame = validate_total_return_strictly_positive(df_total_return_raw)

    # Step 2: Compute discrete simple periodic returns per ticker
    periodic_returns: pd.Series = compute_monthly_returns_per_ticker(df_total_return_valid)

    # Step 3: Create the wide returns matrix aligned to the 18-asset universe
    returns_matrix: pd.DataFrame = construct_wide_returns_matrix(periodic_returns, universe_map, canonical_order)

    # Log the successful completion of the orchestration task
    logger.info("Task 11 completed successfully. Canonical monthly returns panel constructed.")

    # Return the canonical returns matrix
    return returns_matrix
````

````python
# Task 12 — Cleanse macro series and enforce vintage policy

# ==============================================================================
# Task 12: Cleanse macro series and enforce vintage policy
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 12, Step 1: Detect and mark revision-sensitive series (enforce vintage policy)
# -------------------------------------------------------------------------------------------------------------------------------

def enforce_macro_vintage_policy(
    df_macro: pd.DataFrame,
    vintage_policy: str
) -> pd.DataFrame:
    """
    Enforces the point-in-time vintage policy for revision-sensitive macroeconomic
    series (specifically Real GDP Growth).

    CRITICAL REMEDIATION: Replaces the naive row-based `.shift(1)` with a strict
    temporal grid enforcement. The dataframe is reindexed to a continuous monthly
    grid before shifting, guaranteeing a true 1-month calendar lag regardless of
    irregularities or missing rows in the raw data.

    Parameters
    ----------
    df_macro : pd.DataFrame
        The temporally cleansed macro dataframe containing 'real_gdp_growth_rev'.
    vintage_policy : str
        The policy string from the configuration dictating how to handle revisions.

    Returns
    -------
    pd.DataFrame
        The macro dataframe with the vintage policy enforced on revision-sensitive columns.

    Raises
    ------
    TypeError
        If the input is not a pandas DataFrame or lacks a DatetimeIndex.
    ValueError
        If the required 'real_gdp_growth_rev' column is missing.
    """
    # Verify that the input is a pandas DataFrame
    if not isinstance(df_macro, pd.DataFrame):
        raise TypeError(f"df_macro must be a pandas DataFrame, got {type(df_macro).__name__}.")

    # Verify that the dataframe possesses a DatetimeIndex for temporal operations
    if not isinstance(df_macro.index, pd.DatetimeIndex):
        raise TypeError("df_macro must have a DatetimeIndex to enforce temporal vintage policies.")

    # Verify that the revision-sensitive column exists in the dataframe
    if "real_gdp_growth_rev" not in df_macro.columns:
        raise ValueError("df_macro must contain the revision-sensitive column 'real_gdp_growth_rev'.")

    # Create a deep copy to preserve the immutability of the input dataframe
    df_clean: pd.DataFrame = df_macro.copy(deep=True)

    # Check if the policy dictates an approximation due to lack of true ALFRED point-in-time data
    if vintage_policy == "latest_available_with_approximation_disclaimer":

        # Extract the minimum and maximum dates to define the strict temporal boundaries
        min_date: pd.Timestamp = df_clean.index.min()
        max_date: pd.Timestamp = df_clean.index.max()

        # Generate a strict, continuous monthly start (MS) grid
        # This ensures that every single calendar month is represented by exactly one row
        strict_monthly_grid: pd.DatetimeIndex = pd.date_range(start=min_date, end=max_date, freq='MS')

        # Reindex the dataframe to this strict grid. Missing months will be filled with NaN.
        df_reindexed: pd.DataFrame = df_clean.reindex(strict_monthly_grid)

        # Apply a 1-period (1-month) lag to the GDP series on the strict grid
        # Because the grid is strictly monthly, a 1-row shift is mathematically guaranteed
        # to be exactly a 1-month calendar lag.
        df_reindexed["real_gdp_growth_rev"] = df_reindexed["real_gdp_growth_rev"].shift(1)

        # Realign the shifted data back to the original index structure
        # This preserves the original dataframe shape while safely incorporating the lagged data
        df_clean["real_gdp_growth_rev"] = df_reindexed["real_gdp_growth_rev"].reindex(df_clean.index)

        # Log a prominent warning that reproduction is approximate due to the simulated lag
        logger.warning(
            "VINTAGE POLICY ENFORCEMENT: 'real_gdp_growth_rev' has been lagged by exactly 1 calendar month "
            "to simulate publication delay. Reproduction of historical regime calls is approximate due to "
            "the lack of true point-in-time (ALFRED) vintage data."
        )
    else:
        # Log that the data is assumed to be true point-in-time
        logger.info(f"Vintage policy '{vintage_policy}' applied. Assuming data is true point-in-time.")

    # Return the dataframe with the vintage policy enforced
    return df_clean

# -------------------------------------------------------------------------------------------------------------------------------
# Task 12, Step 2: Handle missing macro values via bounded imputation
# -------------------------------------------------------------------------------------------------------------------------------

def apply_bounded_macro_imputation(
    df_macro: pd.DataFrame,
    forward_fill_limit: int
) -> pd.DataFrame:
    """
    Applies a strictly bounded forward-fill imputation to handle missing macroeconomic
    data releases, preventing infinitely stale data from driving regime classifications.

    Parameters
    ----------
    df_macro : pd.DataFrame
        The macro dataframe with vintage policies enforced.
    forward_fill_limit : int
        The maximum number of consecutive periods to forward-fill missing data.

    Returns
    -------
    pd.DataFrame
        The imputed macro dataframe.

    Raises
    ------
    TypeError
        If the input is not a pandas DataFrame or the limit is not an integer.
    ValueError
        If critical indicators still contain NaN values after bounded imputation.
    """
    # Verify that the input is a pandas DataFrame
    if not isinstance(df_macro, pd.DataFrame):
        raise TypeError(f"df_macro must be a pandas DataFrame, got {type(df_macro).__name__}.")

    # Verify that the forward fill limit is an integer
    if not isinstance(forward_fill_limit, int) or forward_fill_limit < 0:
        raise TypeError(f"forward_fill_limit must be a non-negative integer, got {forward_fill_limit}.")

    # Create a deep copy to preserve the immutability of the input dataframe
    df_imputed: pd.DataFrame = df_macro.copy(deep=True)

    # Apply forward-fill imputation with the strict limit defined in the configuration
    # This ensures that if an indicator stops publishing, the pipeline will eventually halt rather than use obsolete data
    df_imputed = df_imputed.ffill(limit=forward_fill_limit)

    # Check for any remaining NaN values post-imputation
    if df_imputed.isna().any().any():
        # Extract the specific columns that still contain missing data
        missing_cols: List[str] = df_imputed.columns[df_imputed.isna().any()].tolist()

        # Raise a fatal ValueError to prevent the Macro Agent from operating on incomplete data
        raise ValueError(
            f"Fatal Data Quality Error: Critical macro indicators still contain NaN values after "
            f"bounded imputation (limit={forward_fill_limit}). Affected columns: {missing_cols}. "
            f"The Macro Agent cannot compute valid z-scores with missing data."
        )

    # Log the successful application of bounded imputation
    logger.info(f"Successfully applied bounded forward-fill imputation (limit={forward_fill_limit}). No NaNs remain.")

    # Return the fully imputed dataframe
    return df_imputed

# -------------------------------------------------------------------------------------------------------------------------------
# Task 12, Step 3: Emit a macro quality report
# -------------------------------------------------------------------------------------------------------------------------------

def generate_macro_quality_report(
    df_macro_clean: pd.DataFrame,
    df_macro_raw: pd.DataFrame,
    output_dir: Path
) -> Dict[str, Any]:
    """
    Generates a comprehensive data quality report for the macro series, quantifying
    missingness ratios, identifying critical trigger events (e.g., negative payrolls),
    and flagging extreme standard deviation moves (e.g., oil shocks).

    Parameters
    ----------
    df_macro_clean : pd.DataFrame
        The fully cleansed and imputed macro dataframe.
    df_macro_raw : pd.DataFrame
        The raw macro dataframe (pre-imputation) to calculate true missingness.
    output_dir : Path
        The directory where the quality report JSON will be saved.

    Returns
    -------
    Dict[str, Any]
        A dictionary containing the structured quality metrics and trigger flags.

    Raises
    ------
    OSError
        If the quality report cannot be written to the specified directory.
    """
    # Initialize the master dictionary to hold the quality report
    quality_report: Dict[str, Any] = {}

    # Calculate the total number of observations in the dataset
    total_obs: int = len(df_macro_raw)

    # 1. Compute missingness ratio per column using the raw (pre-imputation) dataframe
    missingness: Dict[str, float] = (df_macro_raw.isna().sum() / total_obs).to_dict()
    quality_report["missingness_ratios"] = missingness

    # 2. Flag instances of negative nonfarm payrolls (critical recession trigger)
    if "nonfarm_payrolls_raw" in df_macro_clean.columns:
        # Create a boolean mask for negative payroll prints
        negative_payrolls_mask: pd.Series = df_macro_clean["nonfarm_payrolls_raw"] < 0
        # Extract the dates where payrolls were negative
        negative_dates: List[str] = df_macro_clean.index[negative_payrolls_mask].strftime("%Y-%m-%d").tolist()

        quality_report["trigger_events"] = {
            "negative_nonfarm_payrolls_count": int(negative_payrolls_mask.sum()),
            "negative_nonfarm_payrolls_dates": negative_dates
        }

    # 3. Flag extreme standard deviation moves in Brent crude (oil shocks)
    if "brent_crude_usd" in df_macro_clean.columns:
        # Extract the Brent crude series
        brent_series: pd.Series = df_macro_clean["brent_crude_usd"]

        # Calculate the rolling 12-month mean and standard deviation
        rolling_mean: pd.Series = brent_series.rolling(window=12, min_periods=1).mean()
        rolling_std: pd.Series = brent_series.rolling(window=12, min_periods=1).std()

        # Avoid division by zero by replacing 0 std with a small epsilon
        rolling_std = rolling_std.replace(0, 1e-8)

        # Calculate the rolling z-score
        brent_zscore: pd.Series = (brent_series - rolling_mean) / rolling_std

        # Create a boolean mask for extreme moves (z-score > 2.0)
        oil_shock_mask: pd.Series = brent_zscore > 2.0
        # Extract the dates of the oil shocks
        oil_shock_dates: List[str] = df_macro_clean.index[oil_shock_mask].strftime("%Y-%m-%d").tolist()

        # Add to the trigger events dictionary
        if "trigger_events" not in quality_report:
            quality_report["trigger_events"] = {}

        quality_report["trigger_events"]["brent_crude_shocks_count"] = int(oil_shock_mask.sum())
        quality_report["trigger_events"]["brent_crude_shocks_dates"] = oil_shock_dates

    # Ensure the output directory exists, creating it if necessary
    output_dir.mkdir(parents=True, exist_ok=True)

    # Construct the full file path for the quality report JSON
    report_path: Path = output_dir / "macro_quality_report.json"

    # Attempt to write the quality report to disk
    try:
        # Open the file in write mode with UTF-8 encoding
        with open(report_path, "w", encoding="utf-8") as f:
            # Serialize the dictionary to a formatted JSON string and write it
            json.dump(quality_report, f, indent=4)
    except OSError as exc:
        # Raise an OSError if filesystem permissions prevent writing
        raise OSError(f"Failed to write macro quality report to {report_path}: {exc}") from exc

    # Log the successful generation and persistence of the quality report
    logger.info(f"Macro quality report generated and saved to {report_path}.")

    # Return the quality report dictionary
    return quality_report

# -------------------------------------------------------------------------------------------------------------------------------
# Task 12, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_12_cleanse_macro_series(
    df_macro_raw: pd.DataFrame,
    config: Dict[str, Any],
    run_output_dir: Path
) -> pd.DataFrame:
    """
    Orchestrates the cleansing of the macro series, enforcing vintage policies,
    applying bounded imputation, and generating a data quality audit report.

    Parameters
    ----------
    df_macro_raw : pd.DataFrame
        The temporally cleansed raw macro dataframe (post Task 10).
    config : Dict[str, Any]
        The validated study configuration dictionary containing missing data policies.
    run_output_dir : Path
        The root directory for the current run's artifacts.

    Returns
    -------
    pd.DataFrame
        The fully cleansed, imputed, and point-in-time compliant macro dataframe.

    Raises
    ------
    KeyError
        If required configuration keys are missing.
    """
    # Log the commencement of the macro cleansing task
    logger.info("Starting Task 12: Cleanse macro series and enforce vintage policy.")

    # Extract the missing data policy and vintage policy from the configuration
    try:
        missing_data_policy: Dict[str, Any] = config["DATA_CONVENTIONS"]["missing_data_policy"]
        forward_fill_limit: int = missing_data_policy["forward_fill_limit"]

        point_in_time_policy: Dict[str, Any] = config["DATA_CONVENTIONS"]["point_in_time_policy"]
        vintage_policy: str = point_in_time_policy["gdp_vintage_policy"]
    except KeyError as exc:
        # Raise a KeyError if the expected nested structures are missing
        raise KeyError(f"Configuration is missing required keys for Task 12: {exc}") from exc

    # Step 1: Detect and mark revision-sensitive series to enforce the vintage policy
    df_macro_vintage: pd.DataFrame = enforce_macro_vintage_policy(df_macro_raw, vintage_policy)

    # Step 2: Handle missing macro values via bounded imputation
    df_macro_clean: pd.DataFrame = apply_bounded_macro_imputation(df_macro_vintage, forward_fill_limit)

    # Step 3: Emit a macro quality report to quantify missingness and trigger events
    generate_macro_quality_report(df_macro_clean, df_macro_raw, run_output_dir)

    # Log the successful completion of the orchestration task
    logger.info("Task 12 completed successfully. Macro series cleansed and vintage policy enforced.")

    # Return the fully cleansed macro dataframe
    return df_macro_clean
````

````python
# Task 13 — Cleanse equity fundamentals and signals (valuation + technical context)

# ================================================================================
# Task 13: Cleanse equity fundamentals and signals (valuation + technical context)
# ================================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 13, Step 1: Validate equity-only fields to prevent category contamination
# -------------------------------------------------------------------------------------------------------------------------------

def enforce_equity_only_fields(
    df_fundamentals: pd.DataFrame,
    equity_only_cols: List[str]
) -> pd.DataFrame:
    """
    Enforces domain validity by nullifying equity-specific valuation metrics
    (e.g., CAPE, P/E) for non-equity asset classes, preventing the LLM from
    hallucinating invalid valuation logic.

    Parameters
    ----------
    df_fundamentals : pd.DataFrame
        The raw fundamentals dataframe with a MultiIndex containing 'investment_universe'.
    equity_only_cols : List[str]
        The list of column names that are strictly applicable only to equities.

    Returns
    -------
    pd.DataFrame
        The cleansed dataframe with non-equity rows nullified for the specified columns.

    Raises
    ------
    TypeError
        If the input is not a pandas DataFrame or lacks a MultiIndex.
    ValueError
        If the 'investment_universe' level is missing from the MultiIndex.
    """
    # Verify that the input is a pandas DataFrame
    if not isinstance(df_fundamentals, pd.DataFrame):
        raise TypeError(f"df_fundamentals must be a pandas DataFrame, got {type(df_fundamentals).__name__}.")

    # Verify that the dataframe possesses a MultiIndex
    if not isinstance(df_fundamentals.index, pd.MultiIndex):
        raise TypeError("df_fundamentals must have a MultiIndex.")

    # Verify that the 'investment_universe' level exists in the MultiIndex
    if "investment_universe" not in df_fundamentals.index.names:
        raise ValueError("MultiIndex must contain the level 'investment_universe'.")

    # Create a deep copy to preserve the immutability of the input dataframe
    df_clean: pd.DataFrame = df_fundamentals.copy(deep=True)

    # Extract the 'investment_universe' level values for boolean masking
    universe_level: pd.Index = df_clean.index.get_level_values("investment_universe")

    # Create a boolean mask identifying rows that do NOT belong to the 'Equity' category
    # We use str.contains to be robust against slight naming variations (e.g., 'US Equity')
    non_equity_mask: np.ndarray = ~universe_level.str.contains("Equity", case=False, na=False)

    # Identify which of the specified equity-only columns actually exist in the dataframe
    cols_to_nullify: List[str] = [col for col in equity_only_cols if col in df_clean.columns]

    # Apply the mask to nullify the equity-only fields for non-equity assets
    if cols_to_nullify and non_equity_mask.any():
        df_clean.loc[non_equity_mask, cols_to_nullify] = np.nan

        # Log the enforcement action for the audit trail
        logger.info(f"Nullified equity-only fields {cols_to_nullify} for {non_equity_mask.sum()} non-equity rows.")

    # Post-transformation assertion: verify that all non-equity rows have NaNs in these columns
    if cols_to_nullify:
        # Extract the subset of data that should have been nullified
        nullified_subset: pd.DataFrame = df_clean.loc[non_equity_mask, cols_to_nullify]
        # Assert that all values in this subset are indeed NaN
        assert nullified_subset.isna().all().all(), "Fatal Error: Failed to nullify all non-equity rows."

    # Return the cleansed dataframe
    return df_clean

# -------------------------------------------------------------------------------------------------------------------------------
# Task 13, Step 2: Standardize scale conventions to enforce strict unit definitions
# -------------------------------------------------------------------------------------------------------------------------------

def validate_yield_scale_conventions(
    df_fundamentals: pd.DataFrame,
    yield_cols: List[str]
) -> pd.DataFrame:
    """
    Validates that all yield fields are represented as decimals (e.g., 0.05)
    rather than percentages (e.g., 5.0), preventing catastrophic order-of-magnitude
    errors in downstream Capital Market Assumptions (CMAs).

    Parameters
    ----------
    df_fundamentals : pd.DataFrame
        The fundamentals dataframe (post equity-only enforcement).
    yield_cols : List[str]
        The list of column names representing yield metrics.

    Returns
    -------
    pd.DataFrame
        The validated dataframe (returned unchanged if validation passes).

    Raises
    ------
    TypeError
        If the input is not a pandas DataFrame.
    ValueError
        If any yield value strictly exceeds 1.0, indicating a likely percentage scale error.
    """
    # Verify that the input is a pandas DataFrame
    if not isinstance(df_fundamentals, pd.DataFrame):
        raise TypeError(f"df_fundamentals must be a pandas DataFrame, got {type(df_fundamentals).__name__}.")

    # Identify which of the specified yield columns actually exist in the dataframe
    cols_to_check: List[str] = [col for col in yield_cols if col in df_fundamentals.columns]

    # Iterate through each yield column to validate its scale
    for col in cols_to_check:
        # Extract the maximum value in the column, ignoring NaNs
        max_val: float = float(df_fundamentals[col].max(skipna=True))

        # Check if the maximum value exceeds 1.0 (100%)
        # While extreme anomalies exist, a yield > 100% in a broad index is almost certainly a scale error
        if max_val > 1.0:
            # Raise a fatal ValueError forcing upstream data correction
            raise ValueError(
                f"Fatal Scale Error: Column '{col}' contains a value of {max_val:.4f}, which exceeds 1.0. "
                f"Yield fields must be stored in decimal form (e.g., 0.03 = 3%), not percent form (e.g., 3.0 = 3%)."
            )

    # Log the successful validation of the scale conventions
    logger.info(f"Successfully validated decimal scale conventions for yield columns: {cols_to_check}.")

    # Return the validated dataframe
    return df_fundamentals

# -------------------------------------------------------------------------------------------------------------------------------
# Task 13, Step 3: Align fundamentals and signals timestamps to synchronize the feature space
# -------------------------------------------------------------------------------------------------------------------------------

def align_features_to_macro_calendar(
    df_features: pd.DataFrame,
    df_macro_clean: pd.DataFrame,
    df_name: str,
    resample_rule: str = "BME"
) -> pd.DataFrame:
    """
    Synchronizes the temporal index of a feature dataframe (fundamentals or signals)
    to match the canonical macro calendar, ensuring perfect temporal alignment
    across all feature sets for coherent multi-agent reasoning.

    Parameters
    ----------
    df_features : pd.DataFrame
        The feature dataframe (fundamentals or signals) with a MultiIndex.
    df_macro_clean : pd.DataFrame
        The cleansed macro dataframe serving as the canonical temporal grid.
    df_name : str
        The string identifier of the feature dataframe for logging.
    resample_rule : str, optional
        The pandas frequency alias for resampling. Defaults to "BME" (Business Month End).

    Returns
    -------
    pd.DataFrame
        The temporally aligned feature dataframe.

    Raises
    ------
    TypeError
        If inputs are not pandas DataFrames or lack the required index structures.
    ValueError
        If the temporal alignment fails to match the macro calendar.
    """
    # Verify that the inputs are pandas DataFrames
    if not isinstance(df_features, pd.DataFrame):
        raise TypeError(f"df_features must be a pandas DataFrame, got {type(df_features).__name__}.")
    if not isinstance(df_macro_clean, pd.DataFrame):
        raise TypeError(f"df_macro_clean must be a pandas DataFrame, got {type(df_macro_clean).__name__}.")

    # Verify that the feature dataframe possesses a MultiIndex
    if not isinstance(df_features.index, pd.MultiIndex):
        raise TypeError(f"{df_name} must have a MultiIndex.")

    # Verify that the macro dataframe possesses a DatetimeIndex
    if not isinstance(df_macro_clean.index, pd.DatetimeIndex):
        raise TypeError("df_macro_clean must have a DatetimeIndex.")

    # Create a deep copy to preserve the immutability of the input dataframe
    df_aligned: pd.DataFrame = df_features.copy(deep=True)

    # Extract the names of the cross-sectional levels (everything except 'date')
    cross_sectional_levels: List[str] = [name for name in df_aligned.index.names if name != "date"]

    # Group by the cross-sectional levels and resample the 'date' level to the target frequency
    # We use .last() to take the most recent observation within each month
    df_aligned = (
        df_aligned
        .groupby(cross_sectional_levels, group_keys=False)
        .resample(resample_rule, level="date")
        .last()
    )

    # Extract the unique dates from the aligned feature dataframe
    aligned_dates: pd.DatetimeIndex = pd.DatetimeIndex(df_aligned.index.get_level_values("date").unique())

    # Extract the unique dates from the canonical macro dataframe
    macro_dates: pd.DatetimeIndex = pd.DatetimeIndex(df_macro_clean.index.unique())

    # Verify that the aligned dates are a subset of the macro dates (or exactly equal)
    # This ensures no feature data exists on dates where macro data is absent
    unmatched_dates: pd.DatetimeIndex = aligned_dates.difference(macro_dates)

    if not unmatched_dates.empty:
        logger.warning(
            f"Temporal Alignment Warning: {df_name} contains dates not present in the macro calendar: "
            f"{unmatched_dates.strftime('%Y-%m-%d').tolist()}. These will be dropped to ensure perfect alignment."
        )
        # Hard-slice the aligned dataframe to keep only dates present in the macro calendar
        # We use an inner join approach via index intersection
        valid_index_mask: np.ndarray = df_aligned.index.get_level_values("date").isin(macro_dates)
        df_aligned = df_aligned.loc[valid_index_mask]

    # Log the successful temporal alignment
    logger.info(f"Successfully aligned {df_name} timestamps to the canonical macro calendar using rule '{resample_rule}'.")

    # Return the temporally aligned dataframe
    return df_aligned

# -------------------------------------------------------------------------------------------------------------------------------
# Task 13, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_13_cleanse_fundamentals_and_signals(
    df_fundamentals_raw: pd.DataFrame,
    df_signals_raw: pd.DataFrame,
    df_macro_clean: pd.DataFrame,
    config: Dict[str, Any]
) -> Tuple[pd.DataFrame, pd.DataFrame]:
    """
    Orchestrates the cleansing of equity fundamentals and signals, enforcing domain
    validity, standardizing scale conventions, and synchronizing the temporal grid.

    Parameters
    ----------
    df_fundamentals_raw : pd.DataFrame
        The raw fundamentals dataframe (post Task 10 temporal normalization).
    df_signals_raw : pd.DataFrame
        The raw signals dataframe (post Task 10 temporal normalization).
    df_macro_clean : pd.DataFrame
        The fully cleansed macro dataframe (post Task 12) serving as the temporal anchor.
    config : Dict[str, Any]
        The validated study configuration dictionary containing schema definitions.

    Returns
    -------
    Tuple[pd.DataFrame, pd.DataFrame]
        A tuple containing the cleansed and aligned fundamentals and signals dataframes.

    Raises
    ------
    KeyError
        If required configuration keys are missing.
    """
    # Log the commencement of the fundamentals and signals cleansing task
    logger.info("Starting Task 13: Cleanse equity fundamentals and signals (valuation + technical context).")

    # Extract the schema definitions from the configuration
    try:
        fundamentals_schema: Dict[str, Any] = config["RAW_DATA_SCHEMAS"]["FUNDAMENTALS"]
        equity_only_cols: List[str] = fundamentals_schema["validation"]["equity_only_fields"]

        # Define the yield columns that must be validated for decimal scale
        # This is derived from the schema keys that contain the word 'yield'
        yield_cols: List[str] = [col for col in fundamentals_schema["schema"].keys() if "yield" in col.lower()]

        # Extract the resampling rule from the data conventions
        resample_rule: str = config["DATA_CONVENTIONS"]["resampling_rules"].get("daily_to_monthly", "BME")
    except KeyError as exc:
        # Raise a KeyError if the expected nested structures are missing
        raise KeyError(f"Configuration is missing required keys for Task 13: {exc}") from exc

    # Step 1: Validate equity-only fields to prevent category contamination
    df_fundamentals_step1: pd.DataFrame = enforce_equity_only_fields(df_fundamentals_raw, equity_only_cols)

    # Step 2: Standardize scale conventions to enforce strict unit definitions
    df_fundamentals_step2: pd.DataFrame = validate_yield_scale_conventions(df_fundamentals_step1, yield_cols)

    # Step 3: Align fundamentals and signals timestamps to synchronize the feature space
    df_fundamentals_clean: pd.DataFrame = align_features_to_macro_calendar(
        df_features=df_fundamentals_step2,
        df_macro_clean=df_macro_clean,
        df_name="df_fundamentals",
        resample_rule=resample_rule
    )

    df_signals_clean: pd.DataFrame = align_features_to_macro_calendar(
        df_features=df_signals_raw,
        df_macro_clean=df_macro_clean,
        df_name="df_signals",
        resample_rule=resample_rule
    )

    # Log the successful completion of the orchestration task
    logger.info("Task 13 completed successfully. Fundamentals and signals are cleansed and temporally aligned.")

    # Return the cleansed and aligned dataframes
    return df_fundamentals_clean, df_signals_clean
````

````python
# Task 14 — Cleanse benchmark and compute benchmark returns

# ==============================================================================
# Task 14: Cleanse benchmark and compute benchmark returns
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 14, Step 1: Validate benchmark leg indices are total return indices
# -------------------------------------------------------------------------------------------------------------------------------

def validate_benchmark_indices_strictly_positive(df_benchmark: pd.DataFrame) -> pd.DataFrame:
    """
    Validates that the benchmark leg indices contain strictly positive values,
    enforcing the requirement that they represent total return indices rather
    than price indices, which could contain zero or negative values in anomalous
    data feeds.

    Parameters
    ----------
    df_benchmark : pd.DataFrame
        The temporally cleansed benchmark dataframe containing the equity and bond leg indices.

    Returns
    -------
    pd.DataFrame
        The validated dataframe, guaranteed to contain only strictly positive index values.

    Raises
    ------
    TypeError
        If the input is not a pandas DataFrame.
    ValueError
        If the required index columns are missing or contain values <= 0.
    """
    # Verify that the input is a pandas DataFrame
    if not isinstance(df_benchmark, pd.DataFrame):
        raise TypeError(f"df_benchmark must be a pandas DataFrame, got {type(df_benchmark).__name__}.")

    # Define the required index columns for the 60/40 benchmark
    required_cols: Tuple[str, str] = ("equity_leg_total_return_index", "bond_leg_total_return_index")

    # Verify that the required columns exist in the dataframe
    for col in required_cols:
        if col not in df_benchmark.columns:
            raise ValueError(f"df_benchmark must contain the column '{col}'.")

    # Evaluate both series to find any values that are less than or equal to zero
    for col in required_cols:
        non_positive_mask: pd.Series = df_benchmark[col] <= 0

        # Check if any non-positive values exist
        if non_positive_mask.any():
            # Extract the specific index locations of the invalid data for error reporting
            invalid_entries: pd.Index = df_benchmark[non_positive_mask].index
            # Raise a fatal ValueError detailing the exact invalid entries
            raise ValueError(
                f"Fatal Data Integrity Error: '{col}' contains non-positive values. "
                f"Benchmark indices must be strictly positive total return indices. "
                f"Invalid entries found at: {invalid_entries.tolist()}"
            )

    # Log the successful validation of the strictly positive invariant
    logger.info("Successfully validated that all benchmark leg indices are strictly positive.")

    # Return the validated dataframe
    return df_benchmark

# -------------------------------------------------------------------------------------------------------------------------------
# Task 14, Step 2: Compute benchmark leg returns and the combined 60/40 benchmark return
# -------------------------------------------------------------------------------------------------------------------------------

def compute_combined_benchmark_return(df_benchmark_valid: pd.DataFrame) -> pd.Series:
    """
    Computes the combined 60/40 benchmark return series using the frozen
    monthly-rebalanced weighted arithmetic sum formula.

    Equation: r^{(b)}_t = w_{eq} * r^{(eq)}_t + w_{bond} * r^{(bond)}_t
    where r_t = (TR_t / TR_{t-1}) - 1

    Parameters
    ----------
    df_benchmark_valid : pd.DataFrame
        The validated benchmark DataFrame containing strictly positive indices and weights.

    Returns
    -------
    pd.Series
        The computed monthly benchmark return series.

    Raises
    ------
    ValueError
        If the benchmark weights do not sum to 1.0 within a strict tolerance.
    """
    # Extract the static weights from the first row (assuming constant target weights)
    w_equity: float = float(df_benchmark_valid["w_equity"].iloc[0])
    w_bond: float = float(df_benchmark_valid["w_bond"].iloc[0])

    # Validate that the weights sum to exactly 1.0 (within floating-point tolerance)
    weight_sum: float = w_equity + w_bond
    if abs(weight_sum - 1.0) > 1e-6:
        # Raise a ValueError if the weights do not represent a complete portfolio
        raise ValueError(f"Benchmark weights must sum to 1.0, got {weight_sum:.6f} (w_equity={w_equity}, w_bond={w_bond}).")

    # Extract the total return index series for both legs
    tri_eq: pd.Series = df_benchmark_valid["equity_leg_total_return_index"]
    tri_bond: pd.Series = df_benchmark_valid["bond_leg_total_return_index"]

    # Compute the simple periodic returns for the equity leg: r_t = (TR_t / TR_{t-1}) - 1
    # Explicitly disable forward-filling to preserve true volatility
    r_eq: pd.Series = tri_eq.pct_change(fill_method=None)

    # Compute the simple periodic returns for the bond leg: r_t = (TR_t / TR_{t-1}) - 1
    r_bond: pd.Series = tri_bond.pct_change(fill_method=None)

    # Compute the combined benchmark return using the weighted arithmetic sum
    # This simulates a strict monthly rebalancing to the target weights
    r_benchmark: pd.Series = (w_equity * r_eq) + (w_bond * r_bond)

    # Drop the first NaN value resulting from the percentage change operation
    r_benchmark = r_benchmark.dropna()

    # Name the series for clarity in downstream operations
    r_benchmark.name = "benchmark_return"

    # Log the successful computation of the benchmark return series
    logger.info(f"Successfully computed combined benchmark return series. Weights: Equity={w_equity:.2f}, Bond={w_bond:.2f}.")

    # Return the computed return series
    return r_benchmark

# -------------------------------------------------------------------------------------------------------------------------------
# Task 14, Step 3: Align benchmark returns to the portfolio return frequency
# -------------------------------------------------------------------------------------------------------------------------------

def align_benchmark_to_portfolio_returns(
    r_benchmark: pd.Series,
    returns_matrix: pd.DataFrame
) -> pd.Series:
    """
    Synchronizes the temporal index of the benchmark return series to exactly match
    the canonical portfolio returns matrix, ensuring perfect alignment for tracking
    error and relative performance diagnostics.

    Parameters
    ----------
    r_benchmark : pd.Series
        The computed monthly benchmark return series.
    returns_matrix : pd.DataFrame
        The canonical T x 18 monthly returns matrix constructed in Task 11.

    Returns
    -------
    pd.Series
        The temporally aligned benchmark return series.

    Raises
    ------
    TypeError
        If inputs are not of the expected pandas types.
    ValueError
        If the alignment results in a length mismatch, indicating missing benchmark data.
    """
    # Verify that the inputs are of the correct pandas types
    if not isinstance(r_benchmark, pd.Series):
        raise TypeError(f"r_benchmark must be a pandas Series, got {type(r_benchmark).__name__}.")
    if not isinstance(returns_matrix, pd.DataFrame):
        raise TypeError(f"returns_matrix must be a pandas DataFrame, got {type(returns_matrix).__name__}.")

    # Extract the canonical DatetimeIndex from the returns matrix
    canonical_index: pd.DatetimeIndex = pd.DatetimeIndex(returns_matrix.index)

    # Perform an inner join to align the benchmark series to the canonical index
    # We use .reindex with an intersection to ensure we only keep dates present in both
    aligned_index: pd.DatetimeIndex = r_benchmark.index.intersection(canonical_index)

    # Reindex the benchmark series to the aligned index
    r_benchmark_aligned: pd.Series = r_benchmark.loc[aligned_index]

    # Verify that the length of the aligned benchmark series exactly matches the returns matrix
    if len(r_benchmark_aligned) != len(canonical_index):
        # Calculate the number of missing observations
        missing_obs: int = len(canonical_index) - len(r_benchmark_aligned)
        # Raise a fatal ValueError if the benchmark is missing data required for the backtest
        raise ValueError(
            f"Fatal Temporal Alignment Error: Benchmark return series is missing {missing_obs} observations "
            f"present in the canonical returns matrix. Perfect alignment is required for tracking error diagnostics."
        )

    # Log the successful temporal alignment
    logger.info(f"Successfully aligned benchmark returns to portfolio returns matrix. Length: {len(r_benchmark_aligned)}.")

    # Return the temporally aligned benchmark series
    return r_benchmark_aligned

# -------------------------------------------------------------------------------------------------------------------------------
# Task 14, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_14_cleanse_and_compute_benchmark(
    df_benchmark_60_40_raw: pd.DataFrame,
    returns_matrix: pd.DataFrame,
    config: Dict[str, Any]
) -> pd.Series:
    """
    Orchestrates the cleansing of the benchmark data, the computation of the combined
    60/40 return series, and its temporal alignment to the canonical portfolio returns matrix.

    Parameters
    ----------
    df_benchmark_60_40_raw : pd.DataFrame
        The temporally cleansed raw benchmark dataframe (post Task 10).
    returns_matrix : pd.DataFrame
        The canonical T x 18 monthly returns matrix (post Task 11).
    config : Dict[str, Any]
        The validated study configuration dictionary.

    Returns
    -------
    pd.Series
        The fully cleansed, computed, and aligned benchmark return series.

    Raises
    ------
    KeyError
        If required configuration keys are missing.
    """
    # Log the commencement of the benchmark cleansing and computation task
    logger.info("Starting Task 14: Cleanse benchmark and compute benchmark returns.")

    # Step 1: Validate that the benchmark leg indices are strictly positive total return indices
    df_benchmark_valid: pd.DataFrame = validate_benchmark_indices_strictly_positive(df_benchmark_60_40_raw)

    # Step 2: Compute benchmark leg returns and the combined 60/40 benchmark return
    r_benchmark_raw: pd.Series = compute_combined_benchmark_return(df_benchmark_valid)

    # Step 3: Align benchmark returns to the portfolio return frequency
    r_benchmark_aligned: pd.Series = align_benchmark_to_portfolio_returns(r_benchmark_raw, returns_matrix)

    # Log the successful completion of the orchestration task
    logger.info("Task 14 completed successfully. Benchmark return series is computed and aligned.")

    # Return the aligned benchmark return series
    return r_benchmark_aligned
````

````python
# Task 15 — Validate universe_map prevents silent universe drift

# ==============================================================================
# Task 15: Validate `universe_map` prevents silent universe drift
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 15, Step 1: Confirm exactly 18 keys matching the IPS
# -------------------------------------------------------------------------------------------------------------------------------

def validate_universe_map_keys(
    universe_map: Dict[str, Dict[str, Any]],
    config: Dict[str, Any]
) -> Set[str]:
    """
    Validates that the keys in the universe_map exactly match the 18 canonical
    asset class names defined in the Investment Policy Statement (IPS).

    Parameters
    ----------
    universe_map : Dict[str, Dict[str, Any]]
        The mapping dictionary linking canonical asset class names to their metadata.
    config : Dict[str, Any]
        The validated study configuration dictionary containing the IPS asset classes.

    Returns
    -------
    Set[str]
        The validated set of 18 canonical asset class names.

    Raises
    ------
    TypeError
        If the inputs are not dictionaries.
    KeyError
        If there is a mismatch between the universe_map keys and the IPS definition.
    ValueError
        If the IPS definition does not contain exactly 18 assets.
    """
    # Verify that the inputs are dictionaries
    if not isinstance(universe_map, dict):
        raise TypeError(f"universe_map must be a dictionary, got {type(universe_map).__name__}.")
    if not isinstance(config, dict):
        raise TypeError(f"config must be a dictionary, got {type(config).__name__}.")

    # Extract the asset classes dictionary from the IPS governance configuration
    try:
        asset_classes_dict: Dict[str, List[str]] = config["IPS_GOVERNANCE"]["ASSET_CLASSES"]
    except KeyError as exc:
        raise KeyError(f"Configuration is missing required IPS_GOVERNANCE sub-keys: {exc}") from exc

    # Flatten the categorized lists into a single canonical set of asset class names
    canonical_ips_assets: Set[str] = set()
    for category, classes_list in asset_classes_dict.items():
        canonical_ips_assets.update(classes_list)

    # Verify that the IPS defines exactly 18 unique asset classes
    if len(canonical_ips_assets) != 18:
        raise ValueError(f"Fatal Governance Error: IPS must define exactly 18 unique asset classes, found {len(canonical_ips_assets)}.")

    # Extract the keys from the provided universe_map
    mapped_assets: Set[str] = set(universe_map.keys())

    # Compute the symmetric difference to identify any mismatches
    missing_from_map: Set[str] = canonical_ips_assets - mapped_assets
    extra_in_map: Set[str] = mapped_assets - canonical_ips_assets

    # Check for strict set equality
    if missing_from_map or extra_in_map:
        error_msg: str = "Fatal Universe Drift Detected: universe_map keys do not exactly match the IPS definition."
        if missing_from_map:
            error_msg += f" Missing from map: {sorted(missing_from_map)}."
        if extra_in_map:
            error_msg += f" Unrecognized extra keys in map: {sorted(extra_in_map)}."
        # Raise a KeyError detailing the exact mismatch to prevent silent universe drift
        raise KeyError(error_msg)

    # Log the successful validation of the universe map keys
    logger.info("Successfully validated that universe_map keys exactly match the 18 IPS asset classes.")

    # Return the validated set of canonical asset names
    return canonical_ips_assets

# -------------------------------------------------------------------------------------------------------------------------------
# Task 15, Step 2: Confirm every mapped ticker exists in the dataframes
# -------------------------------------------------------------------------------------------------------------------------------

def validate_mapped_tickers_exist(
    universe_map: Dict[str, Dict[str, Any]],
    df_total_return_raw: pd.DataFrame,
    df_ohlcv_raw: pd.DataFrame,
    min_required_obs: int = 24
) -> Set[str]:
    """
    Validates that every ticker defined in the universe_map actually exists
    within the indices of the required dataframes AND possesses sufficient
    temporal depth (non-NaN observations) to support covariance estimation.

    CRITICAL REMEDIATION: Replaces the naive `.unique()` existence check with a
    rigorous `.groupby().count()` depth check. Prevents assets with insufficient
    history from crashing the downstream Ledoit-Wolf shrinkage estimator.

    Parameters
    ----------
    universe_map : Dict[str, Dict[str, Any]]
        The mapping dictionary containing the 'ticker' metadata for each asset.
    df_total_return_raw : pd.DataFrame
        The temporally cleansed total return dataframe.
    df_ohlcv_raw : pd.DataFrame
        The temporally cleansed OHLCV pricing dataframe.
    min_required_obs : int, optional
        The minimum number of valid, non-NaN observations required per ticker.
        Defaults to 24 (2 years of monthly data).

    Returns
    -------
    Set[str]
        The validated set of required tickers.

    Raises
    ------
    TypeError
        If the dataframes do not possess a MultiIndex.
    ValueError
        If any mapped ticker is missing entirely, lacks sufficient temporal depth,
        or if the 'ticker' metadata is missing from the universe_map.
    """
    # Verify that the dataframes possess a MultiIndex
    if not isinstance(df_total_return_raw.index, pd.MultiIndex):
        raise TypeError("df_total_return_raw must have a MultiIndex.")
    if not isinstance(df_ohlcv_raw.index, pd.MultiIndex):
        raise TypeError("df_ohlcv_raw must have a MultiIndex.")

    # Extract the set of required tickers from the universe_map
    required_tickers: Set[str] = set()
    for asset_class, meta in universe_map.items():
        ticker: str = meta.get("ticker")
        if not ticker:
            raise ValueError(f"universe_map is missing the 'ticker' key for asset class '{asset_class}'.")
        required_tickers.add(ticker)

    # --- 1. Validate Total Return Data Depth ---
    # Group by ticker and count the number of valid (non-NaN) observations in the total return index
    tr_counts: pd.Series = df_total_return_raw.groupby(level="ticker")["total_return_index"].count()

    # Identify tickers that are completely missing from the index
    available_tr_tickers: Set[str] = set(tr_counts.index)
    missing_from_tr: Set[str] = required_tickers - available_tr_tickers

    if missing_from_tr:
        raise ValueError(
            f"Fatal Data Availability Error: The following required tickers are completely missing "
            f"from df_total_return_raw: {sorted(missing_from_tr)}."
        )

    # Identify tickers that exist but have insufficient temporal depth
    shallow_tr_tickers: List[str] = [
        ticker for ticker in required_tickers
        if tr_counts.get(ticker, 0) < min_required_obs
    ]

    if shallow_tr_tickers:
        raise ValueError(
            f"Fatal Temporal Depth Error: The following tickers in df_total_return_raw have fewer than "
            f"{min_required_obs} valid observations: {sorted(shallow_tr_tickers)}. "
            f"This will cause singular matrix errors during covariance estimation."
        )

    # --- 2. Validate OHLCV Data Depth ---
    # Group by ticker and count the number of valid observations in the 'close' column
    ohlcv_counts: pd.Series = df_ohlcv_raw.groupby(level="ticker")["close"].count()

    # Identify tickers that are completely missing from the index
    available_ohlcv_tickers: Set[str] = set(ohlcv_counts.index)
    missing_from_ohlcv: Set[str] = required_tickers - available_ohlcv_tickers

    if missing_from_ohlcv:
        raise ValueError(
            f"Fatal Data Availability Error: The following required tickers are completely missing "
            f"from df_ohlcv_raw: {sorted(missing_from_ohlcv)}."
        )

    # Identify tickers that exist but have insufficient temporal depth
    shallow_ohlcv_tickers: List[str] = [
        ticker for ticker in required_tickers
        if ohlcv_counts.get(ticker, 0) < min_required_obs
    ]

    if shallow_ohlcv_tickers:
        raise ValueError(
            f"Fatal Temporal Depth Error: The following tickers in df_ohlcv_raw have fewer than "
            f"{min_required_obs} valid observations: {sorted(shallow_ohlcv_tickers)}. "
            f"This will break technical signal generation."
        )

    # Log the successful validation of ticker availability and depth
    logger.info(
        f"Successfully validated that all {len(required_tickers)} mapped tickers exist "
        f"with at least {min_required_obs} valid observations in both dataframes."
    )

    # Return the validated set of required tickers
    return required_tickers

# -------------------------------------------------------------------------------------------------------------------------------
# Task 15, Step 3: Emit a "universe integrity" report
# -------------------------------------------------------------------------------------------------------------------------------

def generate_universe_integrity_report(
    universe_map: Dict[str, Dict[str, Any]],
    canonical_ips_assets: Set[str],
    required_tickers: Set[str],
    output_dir: Path
) -> Dict[str, Any]:
    """
    Generates and persists a structured JSON report documenting the integrity
    of the asset universe mapping, including category assignments and FX flags.

    Parameters
    ----------
    universe_map : Dict[str, Dict[str, Any]]
        The validated mapping dictionary.
    canonical_ips_assets : Set[str]
        The validated set of 18 canonical asset class names.
    required_tickers : Set[str]
        The validated set of required tickers.
    output_dir : Path
        The directory where the integrity report JSON will be saved.

    Returns
    -------
    Dict[str, Any]
        A dictionary containing the structured integrity metrics.

    Raises
    ------
    OSError
        If the integrity report cannot be written to the specified directory.
    """
    # Initialize the master dictionary to hold the integrity report
    integrity_report: Dict[str, Any] = {
        "status": "VALIDATED",
        "total_asset_classes": len(canonical_ips_assets),
        "total_unique_tickers": len(required_tickers),
        "asset_details": {}
    }

    # Iterate through the universe map to compile metadata for the report
    for asset_class, meta in universe_map.items():
        # Extract category, defaulting to 'Unknown' if missing
        category: str = meta.get("category", "Unknown")

        # Extract FX hedging flag, defaulting to False if missing
        fx_hedged: bool = meta.get("fx_hedged", False)

        # Populate the asset details dictionary
        integrity_report["asset_details"][asset_class] = {
            "ticker": meta.get("ticker"),
            "category": category,
            "fx_hedged": fx_hedged
        }

        # Log a warning if the category is missing, as it is required for downstream logic
        if category == "Unknown":
            logger.warning(f"Universe Integrity Warning: Asset class '{asset_class}' is missing a 'category' definition.")

    # Ensure the output directory exists, creating it if necessary
    output_dir.mkdir(parents=True, exist_ok=True)

    # Construct the full file path for the integrity report JSON
    report_path: Path = output_dir / "universe_integrity_report.json"

    # Attempt to write the integrity report to disk
    try:
        # Open the file in write mode with UTF-8 encoding
        with open(report_path, "w", encoding="utf-8") as f:
            # Serialize the dictionary to a formatted JSON string and write it
            json.dump(integrity_report, f, indent=4)
    except OSError as exc:
        # Raise an OSError if filesystem permissions prevent writing
        raise OSError(f"Failed to write universe integrity report to {report_path}: {exc}") from exc

    # Log the successful generation and persistence of the integrity report
    logger.info(f"Universe integrity report generated and saved to {report_path}.")

    # Return the integrity report dictionary
    return integrity_report

# -------------------------------------------------------------------------------------------------------------------------------
# Task 15, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_15_validate_universe_map(
    universe_map: Dict[str, Dict[str, Any]],
    df_total_return_raw: pd.DataFrame,
    df_ohlcv_raw: pd.DataFrame,
    config: Dict[str, Any],
    run_output_dir: Path
) -> Dict[str, Any]:
    """
    Orchestrates the validation of the universe_map, ensuring strict alignment
    with the IPS mandate and verifying the physical existence of all required
    data before agentic reasoning begins.

    CRITICAL REMEDIATION: Extracts the `window_months` parameter from the
    configuration and explicitly passes it to `validate_mapped_tickers_exist`
    as `min_required_obs`. This ensures that the temporal depth validation
    dynamically aligns with the covariance estimation requirements, preventing
    singular matrix errors downstream.

    Parameters
    ----------
    universe_map : Dict[str, Dict[str, Any]]
        The mapping dictionary linking canonical asset class names to their metadata.
    df_total_return_raw : pd.DataFrame
        The temporally cleansed total return dataframe.
    df_ohlcv_raw : pd.DataFrame
        The temporally cleansed OHLCV pricing dataframe.
    config : Dict[str, Any]
        The validated study configuration dictionary.
    run_output_dir : Path
        The root directory for the current run's artifacts.

    Returns
    -------
    Dict[str, Any]
        The generated universe integrity report dictionary.

    Raises
    ------
    KeyError
        If there is a mismatch between the universe_map keys and the IPS definition,
        or if required configuration keys are missing.
    ValueError
        If any mapped ticker is missing from the required dataframes or lacks
        sufficient temporal depth.
    """
    # Log the commencement of the universe map validation task
    logger.info("Starting Task 15: Validate `universe_map` prevents silent universe drift.")

    # Extract the covariance estimation window from the configuration
    try:
        cov_params: Dict[str, Any] = config["METHODOLOGY_PARAMS"]["COVARIANCE_ESTIMATION"]
        window_months: int = cov_params["window_months"]
    except KeyError as exc:
        # Raise a KeyError if the expected nested structures are missing
        raise KeyError(f"Configuration is missing required keys for Task 15: {exc}") from exc

    # Step 1: Confirm exactly 18 keys matching the IPS
    canonical_ips_assets: Set[str] = validate_universe_map_keys(universe_map, config)

    # Step 2: Confirm every mapped ticker exists in the dataframes with sufficient depth
    # CRITICAL REMEDIATION: Explicitly pass window_months as min_required_obs
    required_tickers: Set[str] = validate_mapped_tickers_exist(
        universe_map=universe_map,
        df_total_return_raw=df_total_return_raw,
        df_ohlcv_raw=df_ohlcv_raw,
        min_required_obs=window_months
    )

    # Step 3: Emit a "universe integrity" report
    integrity_report: Dict[str, Any] = generate_universe_integrity_report(
        universe_map=universe_map,
        canonical_ips_assets=canonical_ips_assets,
        required_tickers=required_tickers,
        output_dir=run_output_dir
    )

    # Log the successful completion of the orchestration task
    logger.info("Task 15 completed successfully. Universe map is validated and drift-free.")

    # Return the integrity report
    return integrity_report
````

````python
# Task 16 — Define and freeze output contracts (JSON schema discipline)

# ==============================================================================
# Custom Exception for Terminal Schema Validation Failures
# ==============================================================================

class SchemaValidationError(Exception):
    """
    Raised when a pipeline artifact fails its terminal schema validation gate.

    This exception signals a fail-closed pipeline halt. It is distinct from
    standard ValueErrors to allow the orchestrator to differentiate between
    data errors and strict machine-to-machine contract violations.
    """
    pass

# ==============================================================================
# Task 16: Define and freeze output contracts (JSON schema discipline)
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 16, Step 1 & 2: Define schemas for key artifacts and freeze required fields/bounds
# -------------------------------------------------------------------------------------------------------------------------------

def generate_artifact_schemas() -> Dict[str, Dict[str, Any]]:
    """
    Generates strict JSON Schema (Draft 7) definitions for the 9 core pipeline artifacts.
    Embeds mathematical bounds (e.g., confidence in [0,1], weights length=18) directly
    into the schemas to prevent LLM hallucination and ensure data integrity.

    Returns
    -------
    Dict[str, Dict[str, Any]]
        A dictionary mapping schema identifiers to their JSON Schema definitions.
    """
    schemas: Dict[str, Dict[str, Any]] = {}

    # 1. macro_view.schema.json
    schemas["macro_view"] = {
        "$schema": "http://json-schema.org/draft-07/schema#",
        "title": "Macro View Artifact",
        "type": "object",
        "properties": {
            "regime": {"type": "string", "enum": ["Expansion", "Late-cycle", "Recession", "Recovery"]},
            "scores": {
                "type": "object",
                "properties": {
                    "s_g": {"type": "number", "minimum": -3.0, "maximum": 3.0},
                    "s_pi": {"type": "number", "minimum": -3.0, "maximum": 3.0},
                    "s_pol": {"type": "number", "minimum": -3.0, "maximum": 3.0},
                    "s_fc": {"type": "number", "minimum": -3.0, "maximum": 3.0}
                },
                "required": ["s_g", "s_pi", "s_pol", "s_fc"],
                "additionalProperties": False
            },
            "confidence": {"type": "number", "minimum": 0.0, "maximum": 1.0},
            "rationale": {"type": "string", "minLength": 10},
            "as_of_date": {"type": "string"}
        },
        "required": ["regime", "scores", "confidence", "rationale"],
        "additionalProperties": False
    }

    # 2. cma_methods.schema.json
    schemas["cma_methods"] = {
        "$schema": "http://json-schema.org/draft-07/schema#",
        "title": "CMA Methods Artifact",
        "type": "object",
        "properties": {
            "asset_class": {"type": "string"},
            "n_methods": {"type": "integer", "minimum": 1},
            "method_results": {
                "type": "array",
                "items": {
                    "type": "object",
                    "properties": {
                        "method_id": {"type": "integer"},
                        "estimate": {"type": ["number", "null"]},
                        "confidence": {"type": ["number", "null"], "minimum": 0.0, "maximum": 1.0},
                        "breakdown": {"type": "object"},
                        "rationale": {"type": "string"}
                    },
                    "required": ["method_id", "estimate", "confidence", "breakdown", "rationale"],
                    "additionalProperties": False
                }
            }
        },
        "required": ["asset_class", "n_methods", "method_results"],
        "additionalProperties": False
    }

    # 3. cma.schema.json
    schemas["cma"] = {
        "$schema": "http://json-schema.org/draft-07/schema#",
        "title": "Final CMA Artifact",
        "type": "object",
        "properties": {
            "asset_class": {"type": "string"},
            "final_estimate": {"type": "number"},
            "method_weights": {"type": "object"},
            "rationale": {"type": "string", "minLength": 10},
            "method_range": {
                "type": "object",
                "properties": {
                    "min": {"type": "number"},
                    "max": {"type": "number"}
                },
                "required": ["min", "max"],
                "additionalProperties": False
            },
            "range_constraint_verified": {"type": "boolean"},
            "credit_spread_duration": {"type": ["number", "null"]},
            "sector_concentration": {"type": ["number", "null"]},
            "as_of_date": {"type": "string"}
        },
        "required": ["asset_class", "final_estimate", "method_weights", "rationale", "method_range"],
        "additionalProperties": False
    }

    # 4. pc_weights.schema.json
    schemas["pc_weights"] = {
        "$schema": "http://json-schema.org/draft-07/schema#",
        "title": "PC Weights Artifact",
        "type": "object",
        "properties": {
            "method": {"type": "string"},
            "weights": {
                "type": "array",
                "items": {"type": "number", "minimum": 0.0}, # Enforce long-only
                "minItems": 18,
                "maxItems": 18
            },
            "sum_check": {"type": "number", "minimum": 0.99999, "maximum": 1.00001}, # Enforce budget constraint
            "rationale": {"type": "string", "minLength": 10}
        },
        "required": ["method", "weights", "sum_check", "rationale"],
        "additionalProperties": False
    }

    # 5. cro_report.schema.json
    schemas["cro_report"] = {
        "$schema": "http://json-schema.org/draft-07/schema#",
        "title": "CRO Risk Report Artifact",
        "type": "object",
        "properties": {
            "candidate_id": {"type": "string"},
            "ips_compliant": {"type": "boolean"},
            "metrics": {
                "type": "object",
                "properties": {
                    "ex_ante_vol": {"type": "number", "minimum": 0.0},
                    "tracking_error": {"type": "number", "minimum": 0.0},
                    "backtest_sharpe": {"type": "number"},
                    "max_drawdown": {"type": "number", "maximum": 0.0},
                    "alpha": {"type": "number"},
                    "beta_M": {"type": "number"},
                    "beta_S": {"type": "number"},
                    "beta_H": {"type": "number"},
                    "r_squared": {"type": "number", "minimum": 0.0, "maximum": 1.0}
                },
                "required": ["ex_ante_vol", "tracking_error", "backtest_sharpe", "max_drawdown", "alpha", "beta_M", "beta_S", "beta_H", "r_squared"],
                "additionalProperties": False
            },
            "compliance_flags": {"type": "object"}
        },
        "required": ["candidate_id", "ips_compliant", "metrics", "compliance_flags"],
        "additionalProperties": False
    }

    # 6. vote.schema.json
    schemas["vote"] = {
        "$schema": "http://json-schema.org/draft-07/schema#",
        "title": "Vote Artifact",
        "type": "object",
        "properties": {
            "voter_id": {"type": "string"},
            "top_5": {
                "type": "array",
                "items": {
                    "type": "object",
                    "properties": {
                        "method": {"type": "string"},
                        "score": {"type": "integer", "enum": [1, 2, 3, 4, 5]}
                    },
                    "required": ["method", "score"],
                    "additionalProperties": False
                },
                "minItems": 5,
                "maxItems": 5
            },
            "bottom_flag": {
                "type": ["object", "null"],
                "properties": {
                    "method": {"type": "string"},
                    "penalty": {"type": "integer", "enum": [-2]}
                },
                "required": ["method", "penalty"],
                "additionalProperties": False
            }
        },
        "required": ["voter_id", "top_5"],
        "additionalProperties": False
    }

    # 7. final_weights.schema.json
    schemas["final_weights"] = {
        "$schema": "http://json-schema.org/draft-07/schema#",
        "title": "Final Weights Artifact",
        "type": "object",
        "properties": {
            "weights": {
                "type": "array",
                "items": {"type": "number", "minimum": 0.0},
                "minItems": 18,
                "maxItems": 18
            },
            "asset_class_labels": {
                "type": "array",
                "items": {"type": "string"},
                "minItems": 18,
                "maxItems": 18
            },
            "sum_check": {"type": "number", "minimum": 0.99999, "maximum": 1.00001},
            "selected_ensemble": {"type": "string"},
            "rationale": {"type": "string", "minLength": 10},
            "range_constraint_verified": {"type": "boolean"}
        },
        "required": ["weights", "asset_class_labels", "sum_check", "selected_ensemble", "rationale"],
        "additionalProperties": False
    }

    # 8. signals.schema.json
    schemas["signals"] = {
        "$schema": "http://json-schema.org/draft-07/schema#",
        "title": "Signals Artifact",
        "type": "object",
        "properties": {
            "asset_class": {"type": "string"},
            "ticker": {"type": "string"},
            "category": {"type": "string"},
            "as_of_date": {"type": "string"},
            "rsi_14d": {"type": ["number", "null"]},
            "momentum_12m": {"type": ["number", "null"]},
            "market_breadth_raw": {"type": ["number", "null"]},
            "net_fund_flows": {"type": ["number", "null"]},
            "positioning_score_raw": {"type": ["number", "null"]},
            "dividend_yield": {"type": ["number", "null"], "maximum": 1.0},
            "buyback_yield": {"type": ["number", "null"], "maximum": 1.0},
            "market_cap_usd": {"type": ["number", "null"]},
            "earnings_growth_forecast": {"type": ["number", "null"]},
            "valuation_change_assumption": {"type": ["number", "null"]},
            "cape_ratio": {"type": ["number", "null"]},
            "pe_trailing": {"type": ["number", "null"]},
            "earnings_yield": {"type": ["number", "null"], "maximum": 1.0},
            "omitted_fields": {"type": "array", "items": {"type": "string"}}
        },
        "required": ["asset_class", "ticker", "category", "as_of_date"],
        "additionalProperties": False
    }

    # 9. historical_stats.schema.json
    schemas["historical_stats"] = {
        "$schema": "http://json-schema.org/draft-07/schema#",
        "title": "Historical Stats Artifact",
        "type": "object",
        "properties": {
            "asset_class": {"type": "string"},
            "ticker": {"type": "string"},
            "as_of_date": {"type": "string"},
            "history_start_actual": {"type": "string"},
            "n_observations": {"type": "integer", "minimum": 24},
            "annualised_return": {"type": "number"},
            "annualised_vol": {"type": "number", "minimum": 0.0},
            "max_drawdown": {"type": "number", "maximum": 0.0},
            "sharpe_ratio_unannualised": {"type": "number"},
            "monthly_returns": {"type": "object"},
            "correlation_matrix": {"type": "object"}
        },
        "required": ["asset_class", "annualised_return", "annualised_vol", "max_drawdown", "n_observations"],
        "additionalProperties": False
    }

    logger.info("Successfully generated strict JSON Schema definitions for all 9 core artifacts.")
    return schemas

def persist_schemas(schemas: Dict[str, Dict[str, Any]], schemas_dir: Path) -> None:
    """
    Writes the generated JSON Schema definitions to disk.

    Parameters
    ----------
    schemas : Dict[str, Dict[str, Any]]
        The dictionary of schema definitions.
    schemas_dir : Path
        The directory where the schema files will be saved.

    Raises
    ------
    OSError
        If the files cannot be written due to filesystem permissions.
    """
    # Ensure the schemas directory exists
    schemas_dir.mkdir(parents=True, exist_ok=True)

    for schema_name, schema_def in schemas.items():
        file_path: Path = schemas_dir / f"{schema_name}.schema.json"
        try:
            with open(file_path, "w", encoding="utf-8") as f:
                json.dump(schema_def, f, indent=4)
            logger.debug(f"Persisted schema: {file_path}")
        except OSError as exc:
            raise OSError(f"Failed to write schema {schema_name} to {file_path}: {exc}") from exc

    logger.info(f"Successfully persisted all schemas to {schemas_dir}.")

# -------------------------------------------------------------------------------------------------------------------------------
# Task 16, Step 3: Implement schema validation as a hard gate between pipeline stages
# -------------------------------------------------------------------------------------------------------------------------------

def validate_artifact(artifact_path: Path, schema_name: str, schemas_dir: Path) -> bool:
    """
    Validates a JSON artifact against its corresponding frozen schema.
    Acts as a hard gate between pipeline stages, enforcing fail-closed semantics.

    Parameters
    ----------
    artifact_path : Path
        The path to the JSON artifact file to be validated.
    schema_name : str
        The identifier of the schema to validate against (e.g., 'macro_view').
    schemas_dir : Path
        The directory containing the frozen schema files.

    Returns
    -------
    bool
        True if the artifact passes validation.

    Raises
    ------
    FileNotFoundError
        If the artifact file or the schema file does not exist.
    SchemaValidationError
        If the artifact fails validation against the schema, or if it contains invalid JSON.
    """
    # 1. Verify artifact file exists
    if not artifact_path.exists():
        raise FileNotFoundError(f"Artifact not found: '{artifact_path}'. Pipeline halted (fail-closed).")

    # 2. Verify schema file exists
    schema_path: Path = schemas_dir / f"{schema_name}.schema.json"
    if not schema_path.exists():
        raise FileNotFoundError(f"Schema not found: '{schema_path}'. Cannot validate artifact.")

    # 3. Load the schema
    try:
        with open(schema_path, "r", encoding="utf-8") as f:
            schema_def: Dict[str, Any] = json.load(f)
    except json.JSONDecodeError as exc:
        raise SchemaValidationError(f"Schema file '{schema_path}' contains invalid JSON: {exc}") from exc

    # 4. Load the artifact
    try:
        with open(artifact_path, "r", encoding="utf-8") as f:
            artifact_data: Any = json.load(f)
    except json.JSONDecodeError as exc:
        raise SchemaValidationError(
            f"SCHEMA VALIDATION FAILURE: '{artifact_path}' contains invalid JSON. "
            f"Original error: {exc}. Pipeline halted (fail-closed)."
        ) from exc

    # 5. Apply jsonschema validation
    try:
        jsonschema.validate(instance=artifact_data, schema=schema_def)
    except ValidationError as exc:
        raise SchemaValidationError(
            f"SCHEMA VALIDATION FAILURE: Artifact '{artifact_path}' failed validation against schema '{schema_name}'. "
            f"Violation: {exc.message}. Pipeline halted (fail-closed)."
        ) from exc

    logger.debug(f"Artifact '{artifact_path}' successfully passed schema validation against '{schema_name}'.")
    return True

# -------------------------------------------------------------------------------------------------------------------------------
# Task 16, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_16_define_output_contracts(config: Dict[str, Any]) -> Callable[[Path, str], bool]:
    """
    Orchestrates the definition and freezing of output contracts by generating
    strict JSON schemas and establishing the validation gate function.

    Parameters
    ----------
    config : Dict[str, Any]
        The validated study configuration dictionary.

    Returns
    -------
    Callable[[Path, str], bool]
        A partially applied validation function (`validate_artifact`) bound to the
        correct schemas directory, ready for use by downstream orchestrators.

    Raises
    ------
    KeyError
        If required configuration keys are missing.
    """
    logger.info("Starting Task 16: Define and freeze output contracts (JSON schema discipline).")

    # Extract the schemas directory path from the configuration
    try:
        schemas_dir_str: str = config["REGISTRIES"]["OUTPUT_CONTRACT_SCHEMAS_DIR"]
        schemas_dir: Path = Path(schemas_dir_str).resolve()
    except KeyError as exc:
        raise KeyError(f"Configuration is missing required keys for Task 16: {exc}") from exc

    # Step 1 & 2: Generate the strict JSON Schema definitions with embedded bounds
    schemas: Dict[str, Dict[str, Any]] = generate_artifact_schemas()

    # Persist the schemas to disk
    persist_schemas(schemas, schemas_dir)

    # Step 3: Create a bound validation function that knows where the schemas live
    # This function will be passed to downstream orchestrators to enforce the hard gates
    def bound_validate_artifact(artifact_path: Path, schema_name: str) -> bool:
        return validate_artifact(artifact_path, schema_name, schemas_dir)

    logger.info("Task 16 completed successfully. Output contracts are frozen and validation gate is established.")

    # Return the bound validation function
    return bound_validate_artifact
````

````python
# Task 17 — Construct the Macro Agent package

# ==============================================================================
# Task 17: Construct the Macro Agent package
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 17, Step 1: Create Macro Agent description file using PROMPT_TEMPLATES
# -------------------------------------------------------------------------------------------------------------------------------

def compile_macro_agent_prompt(
    prompt_template: str,
    ips_text_or_path: str,
    as_of_date: str,
    prior_macro_view_path: Optional[str] = None
) -> str:
    """
    Compiles the Macro Agent prompt by injecting runtime parameters into the
    inferred template. Enforces the strict operational boundary prohibiting
    LLM arithmetic.

    Parameters
    ----------
    prompt_template : str
        The raw prompt template string from STUDY_CONFIG["PROMPT_TEMPLATES"]["MACRO_AGENT"]["content"].
    ips_text_or_path : str
        The text or file path of the Investment Policy Statement.
    as_of_date : str
        The ISO-8601 as-of date for the simulation.
    prior_macro_view_path : Optional[str], optional
        The path to the previous period's macro view, if available. Defaults to None.

    Returns
    -------
    str
        The fully compiled prompt string ready for the 'developer' role.

    Raises
    ------
    ValueError
        If the compiled prompt does not contain the mandatory arithmetic prohibition.
    """
    # Handle the optional prior macro view path
    prior_path_str: str = prior_macro_view_path if prior_macro_view_path else "None (First Run)"

    # Define the mapping of tool placeholders to their actual Python function names
    # This ensures the LLM knows exactly which tools to call
    tool_mappings: Dict[str, str] = {
        "PY_FETCH_MACRO_SNAPSHOT": "fetch_macro_snapshot",
        "PY_SCORE_GROWTH_DIMENSION": "score_growth_dimension",
        "PY_SCORE_INFLATION_DIMENSION": "score_inflation_dimension",
        "PY_SCORE_POLICY_DIMENSION": "score_policy_dimension",
        "PY_SCORE_FINANCIAL_CONDITIONS_DIMENSION": "score_financial_conditions_dimension",
        "PY_CLASSIFY_REGIME": "classify_regime",
        "PY_WRITE_MACRO_VIEW_JSON": "write_macro_view_json",
        "PY_WRITE_ANALYSIS_MD": "write_analysis_md",
        # Output paths are placeholders in the prompt; the actual paths are handled by the tools
        "MACRO_VIEW_OUTPUT_PATH": "macro-view.json",
        "ANALYSIS_MD_OUTPUT_PATH": "analysis.md"
    }

    # Inject the runtime parameters and tool mappings into the template
    compiled_prompt: str = prompt_template.format(
        IPS_TEXT_OR_PATH=ips_text_or_path,
        AS_OF_DATE=as_of_date,
        PRIOR_MACRO_VIEW_PATH=prior_path_str,
        **tool_mappings
    )

    # Verify that the mandatory arithmetic prohibition is present in the compiled prompt
    if "Do not perform arithmetic" not in compiled_prompt:
        raise ValueError(
            "Fatal Prompt Engineering Error: The compiled Macro Agent prompt lacks the "
            "explicit prohibition against performing arithmetic. Separation of concerns violated."
        )

    logger.info("Successfully compiled Macro Agent prompt.")
    return compiled_prompt

# -------------------------------------------------------------------------------------------------------------------------------
# Task 17, Step 2: Implement and register Macro Agent script tools
# -------------------------------------------------------------------------------------------------------------------------------

def register_macro_agent_tools(
    df_macro_raw: pd.DataFrame,
    artifact_dir: Path,
    as_of_date: str
) -> Tuple[Dict[str, Callable], List[Dict[str, Any]]]:
    """
    Constructs the tool registry for the Macro Agent, binding the deterministic
    Python callables to their JSON Schema definitions for the OpenAI API.
    Injects large dataframes via closures to prevent context window overflow.

    Parameters
    ----------
    df_macro_raw : pd.DataFrame
        The cleansed macro dataframe.
    artifact_dir : Path
        The directory where output artifacts will be written.
    as_of_date : str
        The ISO-8601 as-of date.

    Returns
    -------
    Tuple[Dict[str, Callable], List[Dict[str, Any]]]
        A tuple containing:
        1. A dictionary mapping tool names to their partially applied Python callables.
        2. A list of JSON Schema dictionaries defining the tools for the LLM.
    """
    # 1. Define the Python callable registry with dependency injection via functools.partial
    # The LLM will only need to provide the arguments that are NOT pre-filled here.
    callable_registry: Dict[str, Callable] = {
        "fetch_macro_snapshot": partial(fetch_macro_snapshot, df_macro_raw=df_macro_raw),
        "score_growth_dimension": partial(score_growth_dimension, df_macro_raw=df_macro_raw),
        "score_inflation_dimension": partial(score_inflation_dimension, df_macro_raw=df_macro_raw),
        "score_policy_dimension": partial(score_policy_dimension, df_macro_raw=df_macro_raw),
        "score_financial_conditions_dimension": partial(score_financial_conditions_dimension, df_macro_raw=df_macro_raw),
        "classify_regime": classify_regime, # No dataframe injection needed
        "write_macro_view_json": partial(write_macro_view_json, artifact_dir=artifact_dir, as_of_date=as_of_date),
        "write_analysis_md": partial(write_analysis_md, artifact_dir=artifact_dir, as_of_date=as_of_date)
    }

    # 2. Define the JSON Schema array for the OpenAI API
    # These schemas describe exactly what arguments the LLM must provide for each tool.
    tool_schemas: List[Dict[str, Any]] = [
        {
            "type": "function",
            "function": {
                "name": "fetch_macro_snapshot",
                "description": "Retrieve the most recent macro indicator snapshot on or before the as-of date.",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "as_of_date": {"type": "string", "description": "ISO-8601 date string (e.g., '2026-03-31')."}
                    },
                    "required": ["as_of_date"]
                }
            }
        },
        {
            "type": "function",
            "function": {
                "name": "score_growth_dimension",
                "description": "Compute the standardised growth dimension score.",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "macro_snapshot": {"type": "object", "description": "Output of fetch_macro_snapshot."}
                    },
                    "required": ["macro_snapshot"]
                }
            }
        },
        {
            "type": "function",
            "function": {
                "name": "score_inflation_dimension",
                "description": "Compute the standardised inflation dimension score.",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "macro_snapshot": {"type": "object", "description": "Output of fetch_macro_snapshot."}
                    },
                    "required": ["macro_snapshot"]
                }
            }
        },
        {
            "type": "function",
            "function": {
                "name": "score_policy_dimension",
                "description": "Compute the standardised monetary policy dimension score.",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "macro_snapshot": {"type": "object", "description": "Output of fetch_macro_snapshot."}
                    },
                    "required": ["macro_snapshot"]
                }
            }
        },
        {
            "type": "function",
            "function": {
                "name": "score_financial_conditions_dimension",
                "description": "Compute the standardised financial conditions dimension score.",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "macro_snapshot": {"type": "object", "description": "Output of fetch_macro_snapshot."}
                    },
                    "required": ["macro_snapshot"]
                }
            }
        },
        {
            "type": "function",
            "function": {
                "name": "classify_regime",
                "description": "Map the four-dimensional score vector to a regime label and confidence.",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "score_vector": {
                            "type": "object",
                            "description": "Dict with keys 's_g', 's_pi', 's_pol', 's_fc' containing the computed scores."
                        }
                    },
                    "required": ["score_vector"]
                }
            }
        },
        {
            "type": "function",
            "function": {
                "name": "write_macro_view_json",
                "description": "Serialise and persist the macro regime view to macro-view.json.",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "regime": {"type": "string", "description": "Macro regime label."},
                        "scores": {"type": "object", "description": "Four-dimensional score vector."},
                        "confidence": {"type": "number", "description": "Softmax confidence of the selected regime."},
                        "rationale": {"type": "string", "description": "Narrative explaining the classification."}
                    },
                    "required": ["regime", "scores", "confidence", "rationale"]
                }
            }
        },
        {
            "type": "function",
            "function": {
                "name": "write_analysis_md",
                "description": "Format and persist the macro regime analysis narrative to analysis.md.",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "regime": {"type": "string", "description": "Macro regime label."},
                        "rationale": {"type": "string", "description": "Narrative explaining the classification."},
                        "scores": {"type": "object", "description": "Optional four-dimensional score vector."},
                        "confidence": {"type": "number", "description": "Optional softmax confidence."}
                    },
                    "required": ["regime", "rationale"]
                }
            }
        }
    ]

    logger.info("Successfully registered 8 Macro Agent tools.")
    return callable_registry, tool_schemas

# -------------------------------------------------------------------------------------------------------------------------------
# Task 17, Step 3: Execute the ReAct loop and bind input/output contracts
# -------------------------------------------------------------------------------------------------------------------------------

def execute_macro_react_loop(
    client: Any, # OpenAI Client
    llm_config: Any, # AgentLLMConfig
    compiled_prompt: str,
    callable_registry: Dict[str, Callable],
    tool_schemas: List[Dict[str, Any]],
    max_iterations: int = 20
) -> Tuple[bool, str, str]:
    """
    Executes the ReAct (Reasoning and Acting) loop for the Macro Agent.
    Manages state accumulation, tool dispatch, and termination conditions.

    Parameters
    ----------
    client : Any
        The authenticated OpenAI client.
    llm_config : Any
        The strictly typed configuration for the agent.
    compiled_prompt : str
        The compiled developer prompt.
    callable_registry : Dict[str, Callable]
        The dictionary mapping tool names to Python functions.
    tool_schemas : List[Dict[str, Any]]
        The JSON Schema array of available tools.
    max_iterations : int, optional
        The hard iteration ceiling to prevent infinite loops. Defaults to 20.

    Returns
    -------
    Tuple[bool, str, str]
        A tuple containing:
        1. Boolean indicating successful clean termination.
        2. Path to the written macro-view.json.
        3. Path to the written analysis.md.

    Raises
    ------
    RuntimeError
        If the iteration ceiling is hit without clean termination.
    """
    # Initialize the conversation history
    messages: List[Dict[str, str]] = [
        {"role": "developer", "content": compiled_prompt},
        {"role": "user", "content": "Begin macro regime analysis as-of the specified date. Execute the workflow strictly in order."}
    ]

    # State trackers for termination conditions
    json_written: bool = False
    md_written: bool = False
    json_path: str = ""
    md_path: str = ""

    iteration: int = 0
    while iteration < max_iterations:
        iteration += 1
        logger.debug(f"Macro Agent ReAct Loop Iteration {iteration}/{max_iterations}")

        # Invoke the LLM
        # Note: In a real environment, this calls invoke_and_extract_agent_response
        # For this implementation, we assume the function exists and returns a ParsedLLMResponse
        response = invoke_and_extract_agent_response(
            client=client,
            config=llm_config,
            messages=messages,
            tools=tool_schemas
        )

        # Append the assistant's response to the history
        # We must reconstruct the message dict as expected by the API
        assistant_msg: Dict[str, Any] = {"role": "assistant"}
        if response.content:
            assistant_msg["content"] = response.content
        if response.tool_calls:
            assistant_msg["tool_calls"] = response.tool_calls
        messages.append(assistant_msg)

        # Check if tools were called
        if response.tool_calls:
            for tool_call in response.tool_calls:
                tool_name: str = tool_call.function.name

                # Parse the JSON arguments provided by the LLM
                try:
                    tool_args: Dict[str, Any] = json.loads(tool_call.function.arguments)
                except json.JSONDecodeError:
                    tool_args = {}
                    logger.error(f"Failed to parse arguments for tool {tool_name}")

                # Dispatch to the Python callable
                if tool_name in callable_registry:
                    try:
                        # Execute the deterministic Python script
                        result: Any = callable_registry[tool_name](**tool_args)

                        # Track termination conditions
                        if tool_name == "write_macro_view_json":
                            json_written = True
                            json_path = str(result)
                        elif tool_name == "write_analysis_md":
                            md_written = True
                            md_path = str(result)

                        # Serialize the result for the LLM
                        result_str: str = json.dumps(result)
                    except Exception as exc:
                        # Catch execution errors and feed them back to the LLM
                        result_str = f"Error executing {tool_name}: {str(exc)}"
                        logger.error(result_str)
                else:
                    result_str = f"Error: Tool {tool_name} not found in registry."

                # Append the tool result to the conversation history
                messages.append({
                    "role": "tool",
                    "tool_call_id": tool_call.id,
                    "name": tool_name,
                    "content": result_str
                })

        # Check termination conditions
        if json_written and md_written:
            logger.info("Macro Agent ReAct loop terminated cleanly. All artifacts written.")
            return True, json_path, md_path

        # If the LLM stopped calling tools but hasn't finished, prompt it to continue
        if not response.tool_calls and not (json_written and md_written):
            messages.append({
                "role": "user",
                "content": "You have not completed the workflow. You must call write_macro_view_json and write_analysis_md before terminating."
            })

    # If the loop exits without returning, the ceiling was hit
    raise RuntimeError(f"Macro Agent ReAct loop hit iteration ceiling ({max_iterations}) without clean termination.")

# -------------------------------------------------------------------------------------------------------------------------------
# Task 17, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_17_macro_agent(
    df_macro_raw: pd.DataFrame,
    config: Dict[str, Any],
    run_output_dir: Path,
    client: Any, # OpenAI Client
    validate_artifact_fn: Callable[[Path, str], bool]
) -> Tuple[str, str]:
    """
    Orchestrates the Macro Agent package, executing the ReAct loop to classify
    the economic regime and applying the hard schema validation gate on the output.

    Parameters
    ----------
    df_macro_raw : pd.DataFrame
        The cleansed macro dataframe.
    config : Dict[str, Any]
        The validated study configuration dictionary.
    run_output_dir : Path
        The root directory for the current run's artifacts.
    client : Any
        The authenticated OpenAI client.
    validate_artifact_fn : Callable[[Path, str], bool]
        The bound validation function from Task 16.

    Returns
    -------
    Tuple[str, str]
        Paths to the generated macro-view.json and analysis.md artifacts.

    Raises
    ------
    KeyError
        If required configuration keys are missing.
    SchemaValidationError
        If the generated artifact fails the hard validation gate.
    """
    logger.info("Starting Task 17: Construct the Macro Agent package.")

    # Extract configuration parameters
    try:
        prompt_template: str = config["PROMPT_TEMPLATES"]["MACRO_AGENT"]["content"]
        as_of_date: str = config["AS_OF_DATE"]
        # In a full implementation, IPS text would be loaded from disk
        ips_text: str = "IPS Constraints: Target CPI+3-4%, Vol 8-12%, TE < 6%, MDD > -25%"

        # Extract LLM configuration
        llm_policy_dict: Dict[str, Any] = config["INFRASTRUCTURE"]["LLM_INVOCATION_POLICY"]
        effort_schedule: Dict[str, Any] = config["INFRASTRUCTURE"]["REASONING_EFFORT_SCHEDULE"]["macro_agent"]

        # Construct the AgentLLMConfig (assuming the dataclass is available)
        llm_config = AgentLLMConfig(
            model_name=llm_policy_dict["model"],
            reasoning_effort=effort_schedule["effort"],
            max_completion_tokens=effort_schedule["max_output_tokens"]
        )

        max_iterations: int = config["INFRASTRUCTURE"]["REACT_LOOP_PARAMS"]["macro_agent"]["max_iterations"]
    except KeyError as exc:
        raise KeyError(f"Configuration is missing required keys for Task 17: {exc}") from exc

    # Define the specific artifact directory for the Macro Agent
    macro_artifact_dir: Path = run_output_dir / "macro"

    # Step 1: Compile the prompt
    compiled_prompt: str = compile_macro_agent_prompt(
        prompt_template=prompt_template,
        ips_text_or_path=ips_text,
        as_of_date=as_of_date
    )

    # Step 2: Register tools
    callable_registry, tool_schemas = register_macro_agent_tools(
        df_macro_raw=df_macro_raw,
        artifact_dir=macro_artifact_dir,
        as_of_date=as_of_date
    )

    # Step 3: Execute ReAct loop
    success, json_path_str, md_path_str = execute_macro_react_loop(
        client=client,
        llm_config=llm_config,
        compiled_prompt=compiled_prompt,
        callable_registry=callable_registry,
        tool_schemas=tool_schemas,
        max_iterations=max_iterations
    )

    # Apply the hard schema validation gate (fail-closed)
    json_path: Path = Path(json_path_str)
    logger.info("Applying terminal schema validation gate to macro-view.json.")
    validate_artifact_fn(json_path, "macro_view")

    logger.info("Task 17 completed successfully. Macro Agent artifacts generated and validated.")

    return json_path_str, md_path_str
````

````python
# Task 18 — Construct AC Agent packages (equity, fixed income, real assets, cash)

# ===============================================================================
# Task 18: Construct AC Agent packages (equity, fixed income, real assets, cash)
# ===============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 18, Step 1: Generate the IPS Registry Markdown and Extract Global Financial Parameters
# -------------------------------------------------------------------------------------------------------------------------------

def generate_ips_markdown(ips_config: Dict[str, Any], artifact_dir: Path) -> str:
    """
    Serializes the Investment Policy Statement (IPS) configuration dictionary into
    a structured Markdown document. This serves as the operational boundary and
    context window constraint for all agents in the pipeline.

    Parameters
    ----------
    ips_config : Dict[str, Any]
        The validated IPS_GOVERNANCE dictionary from the study configuration.
    artifact_dir : Path
        The base directory where the IPS.md artifact will be written.

    Returns
    -------
    str
        The absolute path to the written IPS.md file.

    Raises
    ------
    TypeError
        If the inputs are of incorrect types.
    OSError
        If the file cannot be written due to filesystem permissions.
    """
    # Verify input types
    if not isinstance(ips_config, dict):
        raise TypeError(f"ips_config must be a dictionary, got {type(ips_config).__name__}.")
    if not isinstance(artifact_dir, Path):
        raise TypeError(f"artifact_dir must be a pathlib.Path, got {type(artifact_dir).__name__}.")

    # Initialize the markdown lines array
    md_lines: List[str] = [
        "# Investment Policy Statement (IPS)",
        "",
        "This document serves as the strict operational boundary for all autonomous agents.",
        "",
        "## 1. Investment Universe",
        f"- **Total Asset Classes:** {ips_config.get('ASSET_UNIVERSE_N', 18)}",
    ]

    # Extract and format the asset classes by category
    asset_classes: Dict[str, List[str]] = ips_config.get("ASSET_CLASSES", {})
    for category, classes in asset_classes.items():
        md_lines.append(f"### {category.replace('_', ' ').title()}")
        for ac in classes:
            md_lines.append(f"- {ac}")
    md_lines.append("")

    # Extract and format the objective function constraints
    objectives: Dict[str, Any] = ips_config.get("OBJECTIVE", {})
    md_lines.extend([
        "## 2. Objective Function",
        f"- **Target Real Return:** {objectives.get('TARGET_REAL_RETURN_SPREAD', 'CPI + 3.0-4.0%')}",
        f"- **Expected Volatility Band:** {objectives.get('VOLATILITY_BAND', '(0.08, 0.12)')}",
        f"- **Maximum Drawdown Limit:** {objectives.get('MAX_DRAWDOWN_LIMIT', -0.25)}",
        ""
    ])

    # Extract and format the active risk budget
    active_risk: Dict[str, Any] = ips_config.get("ACTIVE_RISK", {})
    md_lines.extend([
        "## 3. Active Risk Budget",
        f"- **Ex-Ante Tracking Error Limit:** {active_risk.get('TRACKING_ERROR_BUDGET_EX_ANTE', 0.06)}",
        f"- **Benchmark:** {active_risk.get('BENCHMARK_ID', '60/40_equity_bond')}",
        ""
    ])

    # Extract and format implementation constraints
    impl_constraints: Dict[str, Any] = ips_config.get("IMPLEMENTATION_CONSTRAINTS", {})
    md_lines.extend([
        "## 4. Implementation Constraints",
        f"- **Long Only:** {impl_constraints.get('long_only', True)}",
        f"- **Max Weight Per Asset:** {impl_constraints.get('max_weight_per_asset', 0.25)}",
        f"- **Min Weight Per Asset:** {impl_constraints.get('min_weight_per_asset', 0.00)}",
        f"- **Cash Allowed:** {impl_constraints.get('cash_allowed', True)}",
        ""
    ])

    # Join the lines into a single markdown string
    md_content: str = "\n".join(md_lines)

    # Ensure the target directory exists
    ips_dir: Path = artifact_dir / "ips"
    ips_dir.mkdir(parents=True, exist_ok=True)

    # Define the output file path
    output_path: Path = ips_dir / "IPS.md"

    # Write the markdown content to disk
    try:
        with open(output_path, "w", encoding="utf-8") as fh:
            fh.write(md_content)
    except OSError as exc:
        raise OSError(f"Failed to write IPS.md to '{output_path}': {exc}") from exc

    logger.info(f"Successfully generated and persisted IPS.md to {output_path}")
    return str(output_path.resolve())


def extract_current_risk_free_rate(df_benchmark_factors_raw: pd.DataFrame, as_of_date: str) -> float:
    """
    Extracts the most recent daily risk-free rate on or before the as-of date
    and annualizes it using the frozen act/365 convention.

    Equation: r_{f,ann} = (1 + r_{f,daily})^{365} - 1

    Parameters
    ----------
    df_benchmark_factors_raw : pd.DataFrame
        The daily benchmark factors dataframe containing the 'rf' column.
    as_of_date : str
        The ISO-8601 as-of date string.

    Returns
    -------
    float
        The annualized risk-free rate in decimal form.

    Raises
    ------
    ValueError
        If the 'rf' column is missing or no data exists prior to the as-of date.
    """
    # Parse the as-of date
    as_of_ts: pd.Timestamp = pd.Timestamp(as_of_date)

    # Ensure the index is timezone-naive for comparison
    df_clean: pd.DataFrame = df_benchmark_factors_raw.copy()
    if df_clean.index.tz is not None:
        df_clean.index = df_clean.index.tz_localize(None)

    # Apply point-in-time filter
    df_filtered: pd.DataFrame = df_clean.loc[df_clean.index <= as_of_ts]

    if df_filtered.empty or "rf" not in df_filtered.columns:
        raise ValueError(f"Cannot extract risk-free rate: No 'rf' data available on or before {as_of_date}.")

    # Extract the most recent daily risk-free rate, forward-filling if the exact date is missing
    rf_daily_raw: Any = df_filtered["rf"].ffill().iloc[-1]

    if pd.isna(rf_daily_raw):
        raise ValueError(f"The most recent risk-free rate observation before {as_of_date} is NaN.")

    rf_daily: float = float(rf_daily_raw)

    # Annualize using the act/365 convention
    rf_annualized: float = float((1.0 + rf_daily) ** 365 - 1.0)

    logger.info(f"Extracted and annualized risk-free rate: {rf_annualized:.6f} (from daily rate {rf_daily:.6f})")
    return rf_annualized


def extract_market_cap_weights(
    df_fundamentals_raw: pd.DataFrame,
    as_of_date: str,
    canonical_order: List[str],
    universe_map: Dict[str, Dict[str, Any]]
) -> np.ndarray:
    """
    Extracts the market capitalization for the 18 canonical assets as of the
    specified date and normalizes them into a weight vector for the Black-Litterman prior.

    CRITICAL REMEDIATION: Removes the economically absurd global mean fallback.
    If an asset is missing market cap data, it is imputed using the mean of its
    specific IPS category. If the entire category is missing, the pipeline fails-closed,
    preventing the hallucination of economic relationships across incomparable assets.

    Equation: w_{mkt, i} = MC_i / sum_{j=1}^{18} MC_j

    Parameters
    ----------
    df_fundamentals_raw : pd.DataFrame
        The fundamentals dataframe with a MultiIndex ['date', 'ticker', 'investment_universe'].
    as_of_date : str
        The ISO-8601 as-of date string.
    canonical_order : List[str]
        The strict, ordered list of the 18 canonical asset class names.
    universe_map : Dict[str, Dict[str, Any]]
        The mapping dictionary linking canonical asset class names to their metadata,
        including the 'category' required for imputation.

    Returns
    -------
    np.ndarray
        An 18-element numpy array representing the market-cap weights, summing to 1.0.

    Raises
    ------
    TypeError
        If the dataframe does not possess the required MultiIndex.
    ValueError
        If no market cap data is available, if an entire category is missing,
        or if the resulting vector does not have exactly 18 elements or sums to zero.
    """
    # Verify that the dataframe possesses a MultiIndex to allow cross-sectional slicing
    if not isinstance(df_fundamentals_raw.index, pd.MultiIndex):
        raise TypeError("df_fundamentals_raw must have a MultiIndex.")

    # Parse the as-of date string into a pandas Timestamp for temporal filtering
    as_of_ts: pd.Timestamp = pd.Timestamp(as_of_date)

    # Extract the temporal level from the MultiIndex
    temporal_index: pd.DatetimeIndex = df_fundamentals_raw.index.get_level_values("date")

    # Dynamically inspect and align timezones to prevent slicing errors
    if temporal_index.tz is not None:
        as_of_ts = as_of_ts.tz_localize(temporal_index.tz)

    # Apply point-in-time filter: retain only rows on or before the as-of date
    past_and_present_mask: np.ndarray = temporal_index <= as_of_ts
    df_filtered: pd.DataFrame = df_fundamentals_raw.loc[past_and_present_mask].copy()

    # Guard: Ensure data exists after filtering
    if df_filtered.empty or "market_cap_usd" not in df_filtered.columns:
        raise ValueError(f"Cannot extract market caps: No data available on or before {as_of_date}.")

    # Extract the most recent observation for each ticker
    # Group by ticker and take the last valid observation chronologically
    latest_caps_by_ticker: pd.Series = df_filtered.groupby(level='ticker')["market_cap_usd"].last()

    # Create a mapping from canonical asset class name to its latest market cap
    # This bridges the gap between the raw ticker data and the IPS canonical order
    caps_by_asset_class: Dict[str, float] = {}
    for asset_class, meta in universe_map.items():
        ticker: str = meta.get("ticker")
        if ticker in latest_caps_by_ticker.index:
            caps_by_asset_class[asset_class] = latest_caps_by_ticker[ticker]
        else:
            caps_by_asset_class[asset_class] = np.nan

    # Convert the dictionary to a pandas Series aligned to the canonical order
    aligned_caps: pd.Series = pd.Series(caps_by_asset_class).reindex(canonical_order)

    # -----------------------------------------------------------------------
    # Category-Average Imputation Logic (Strict Fail-Closed)
    # -----------------------------------------------------------------------
    # If any market caps are missing, we impute them using the mean market cap
    # of the available assets strictly within the same IPS category.
    if aligned_caps.isna().any():
        logger.info("Missing market caps detected. Applying strict category-average imputation.")

        # Iterate through the canonical order to impute missing values
        for asset_class in canonical_order:
            if pd.isna(aligned_caps[asset_class]):
                # Extract the category for the missing asset
                category: str = universe_map.get(asset_class, {}).get("category", "Unknown")

                # Identify all other assets in the same category
                peers_in_category: List[str] = [
                    ac for ac, meta in universe_map.items()
                    if meta.get("category") == category
                ]

                # Calculate the mean market cap of the peers
                peer_caps: pd.Series = aligned_caps[peers_in_category]
                category_mean: float = float(peer_caps.mean(skipna=True))

                # CRITICAL REMEDIATION: If the category mean is NaN (i.e., all peers are missing),
                # we must fail-closed. Falling back to a global mean across disparate asset classes
                # (e.g., using Equity caps for Fixed Income) is economically invalid.
                if pd.isna(category_mean):
                    raise ValueError(
                        f"Fatal Imputation Error: Cannot compute market cap for '{asset_class}'. "
                        f"All peer assets in category '{category}' are also missing market cap data. "
                        f"Pipeline halted to prevent distortion of the Black-Litterman equilibrium prior."
                    )

                # Impute using the valid category mean
                aligned_caps[asset_class] = category_mean
                logger.debug(f"Imputed market cap for '{asset_class}' (Category: {category}) with value: {category_mean:e}")

    # Convert the fully imputed Series to a NumPy array of floats
    caps_array: np.ndarray = aligned_caps.values.astype(np.float64)

    # -----------------------------------------------------------------------
    # Normalization
    # -----------------------------------------------------------------------
    # Calculate the total market capitalization across the universe
    total_cap: float = float(caps_array.sum())

    # Guard against zero or negative total capitalization
    if total_cap <= 0:
        raise ValueError(f"Total market capitalization is zero or negative ({total_cap}); cannot compute weights.")

    # Normalize the array so that the weights sum exactly to 1.0
    w_mkt: np.ndarray = caps_array / total_cap

    # Verify dimensionality matches the 18-asset universe
    if w_mkt.shape != (18,):
        raise ValueError(f"Market cap weight vector must have exactly 18 elements, got {w_mkt.shape[0]}.")

    # Verify sum-to-one constraint within floating-point tolerance
    if abs(w_mkt.sum() - 1.0) > 1e-6:
        raise ValueError(f"Market cap weight vector must sum to 1.0, got {w_mkt.sum():.6f}.")

    logger.info("Successfully extracted, imputed, and normalized 18-element market-cap weight vector.")

    # Return the normalized weight vector
    return w_mkt

# -------------------------------------------------------------------------------------------------------------------------------
# Task 18, Step 2: Compile AC Agent Prompts
# -------------------------------------------------------------------------------------------------------------------------------

def compile_ac_agent_prompt(
    asset_class_config: Dict[str, Any],
    prompt_templates: Dict[str, Any],
    macro_view_path: str,
    as_of_date: str,
    rf: float,
    ips_path: str
) -> str:
    """
    Compiles the Asset Class (AC) Agent prompt by injecting runtime parameters.
    Strictly respects the "VERBATIM" provenance flag to ensure cryptographic
    fidelity to the manuscript.

    Parameters
    ----------
    asset_class_config : Dict[str, Any]
        The configuration dictionary for the specific asset class.
    prompt_templates : Dict[str, Any]
        The dictionary containing all prompt templates.
    macro_view_path : str
        The path to the macro-view.json artifact.
    as_of_date : str
        The ISO-8601 as-of date for the simulation.
    rf : float
        The current risk-free rate.
    ips_path : str
        The path to the generated IPS.md file.

    Returns
    -------
    str
        The fully compiled prompt string ready for the 'developer' role.

    Raises
    ------
    KeyError
        If the required prompt template is missing.
    """
    # Extract asset class metadata
    asset_class_name: str = asset_class_config.get("slug", "unknown").replace("_", " ").title()
    # Override for exact match with verbatim template if necessary
    if asset_class_config.get("slug") == "us_large_cap":
        asset_class_name = "US Large Cap"

    asset_ticker: str = asset_class_config.get("ticker", "Unknown")
    history_start: str = asset_class_config.get("history_start", "Unknown")
    macro_sensitivity: str = asset_class_config.get("macro_sensitivity", "Unknown")
    template_key: str = asset_class_config.get("prompt_template_key", "")

    # Retrieve the template dictionary
    if template_key not in prompt_templates:
        raise KeyError(f"Prompt template '{template_key}' not found in configuration.")

    template_dict: Dict[str, Any] = prompt_templates[template_key]
    raw_content: str = template_dict.get("content", "")
    provenance: str = template_dict.get("provenance", "")

    # Check provenance: If VERBATIM, return exactly as is without mutation
    if provenance.startswith("VERBATIM"):
        logger.debug(f"Template '{template_key}' is VERBATIM. Returning without mutation.")
        return raw_content

    # For INFERRED templates, construct a comprehensive kwargs dictionary for safe interpolation
    # We include all possible tool mappings and variables to prevent KeyError during .format()
    injection_kwargs: Dict[str, str] = {
        "ASSET_CLASS_NAME": asset_class_name,
        "ASSET_TICKER": asset_ticker,
        "HISTORY_START": history_start,
        "AS_OF_DATE": as_of_date,
        "MACRO_SENSITIVITY_NOTES": macro_sensitivity,
        "MACRO_VIEW_PATH": macro_view_path,
        "RF": str(rf),
        "IPS_TEXT_OR_PATH": ips_path,
        "FI_CURVES_SPREADS_PATH": "df_fixed_income_curves_spreads_raw (in-memory via tool fetch_signals)",
        # Tool mappings
        "PY_FETCH_HISTORICAL_STATS": "fetch_historical_stats",
        "PY_FETCH_SIGNALS": "fetch_signals",
        "PY_RUN_CMA_METHOD_1": "run_cma_method_1",
        "PY_RUN_CMA_METHOD_2": "run_cma_method_2",
        "PY_RUN_CMA_METHOD_3": "run_cma_method_3",
        "PY_RUN_CMA_METHOD_4": "run_cma_method_4",
        "PY_RUN_CMA_METHOD_5": "run_cma_method_5",
        "PY_RUN_CMA_METHOD_6": "run_cma_method_6",
        "PY_RUN_CMA_METHOD_7": "run_cma_method_7",
        "PY_RUN_FI_CMA_BUILDER": "run_fi_cma_builder",
        "PY_FETCH_CASH_YIELD_PROXY": "fetch_cash_yield_proxy",
        "PY_COMPUTE_HISTORICAL_STATS": "fetch_historical_stats",
        "PY_WRITE_CMA_METHODS_JSON": "write_cma_methods_json",
        "PY_WRITE_CMA_JSON": "write_cma_json",
        "PY_WRITE_SIGNALS_JSON": "write_signals_json",
        "PY_WRITE_HISTORICAL_STATS_JSON": "write_historical_stats_json",
        "PY_WRITE_ANALYSIS_MD": "write_analysis_md",
        # Output paths
        "CMA_METHODS_OUTPUT_PATH": "cma_methods.json",
        "CMA_OUTPUT_PATH": "cma.json",
        "SIGNALS_OUTPUT_PATH": "signals.json",
        "HISTORICAL_STATS_OUTPUT_PATH": "historical_stats.json",
        "SCENARIOS_OUTPUT_PATH": "scenarios.json",
        "CORRELATION_ROW_OUTPUT_PATH": "correlation_row.json",
        "ANALYSIS_MD_OUTPUT_PATH": "analysis.md"
    }

    # Perform strict, immutable string formatting
    try:
        compiled_prompt: str = raw_content.format(**injection_kwargs)
    except KeyError as exc:
        # If the template contains a placeholder not in our comprehensive dict, raise an error
        raise KeyError(f"Template '{template_key}' contains an unresolved placeholder: {exc}") from exc

    logger.debug(f"Successfully compiled INFERRED AC Agent prompt for {asset_class_name}.")
    return compiled_prompt

# -------------------------------------------------------------------------------------------------------------------------------
# Task 18, Step 3: Implement and Register AC Agent Script Tools
# -------------------------------------------------------------------------------------------------------------------------------

def register_ac_agent_tools(
    asset_class_name: str,
    category: str,
    df_total_return_raw: pd.DataFrame,
    df_signals_raw: pd.DataFrame,
    df_fundamentals_raw: pd.DataFrame,
    df_survey_cma_raw: pd.DataFrame,
    df_fixed_income_curves_spreads_raw: pd.DataFrame,
    universe_map: Dict[str, Dict[str, Any]],
    macro_view: Dict[str, Any],
    rf: float,
    sigma: np.ndarray,
    w_mkt: np.ndarray,
    artifact_dir: Path,
    as_of_date: str,
    fi_assumptions: Dict[str, Dict[str, float]],
    duration_map: Dict[str, float]
) -> Tuple[Dict[str, Callable], List[Dict[str, Any]]]:
    """
    Constructs the tool registry for an AC Agent, binding the deterministic
    Python callables to their JSON Schema definitions.

    CRITICAL FIX: Added fi_assumptions and duration_map to the signature and
    injected them into the run_fi_cma_builder tool via functools.partial. This
    ensures the LLM cannot hallucinate actuarial assumptions.

    Parameters
    ----------
    asset_class_name : str
        The name of the asset class.
    category : str
        The category of the asset class (Equity, Fixed Income, Real Assets, Cash).
    df_total_return_raw : pd.DataFrame
        The cleansed total return dataframe.
    df_signals_raw : pd.DataFrame
        The cleansed signals dataframe.
    df_fundamentals_raw : pd.DataFrame
        The cleansed fundamentals dataframe.
    df_survey_cma_raw : pd.DataFrame
        The cleansed survey CMA dataframe.
    df_fixed_income_curves_spreads_raw : pd.DataFrame
        The cleansed fixed income curves dataframe.
    universe_map : Dict[str, Dict[str, Any]]
        The mapping dictionary linking canonical asset class names to their metadata.
    macro_view : Dict[str, Any]
        The parsed macro-view.json dictionary.
    rf : float
        The current annualized risk-free rate.
    sigma : np.ndarray
        The 18x18 Ledoit-Wolf covariance matrix.
    w_mkt : np.ndarray
        The 18-element market capitalization weight vector.
    artifact_dir : Path
        The base directory where output artifacts will be written.
    as_of_date : str
        The ISO-8601 as-of date.
    fi_assumptions : Dict[str, Dict[str, float]]
        Configuration dictionary for FI assumptions.
    duration_map : Dict[str, float]
        Dictionary mapping asset classes to their explicit credit spread duration.

    Returns
    -------
    Tuple[Dict[str, Callable], List[Dict[str, Any]]]
        A tuple containing the callable registry and the JSON Schema array.
    """
    callable_registry: Dict[str, Callable] = {}
    tool_schemas: List[Dict[str, Any]] = []

    # ------------------------------------------------------------------
    # Base tools applicable to Equity, Real Assets, and Cash (partially)
    # ------------------------------------------------------------------
    if category in ["Equity", "Real Assets", "Fixed Income", "Cash"]:
        callable_registry["fetch_historical_stats"] = partial(
            fetch_historical_stats,
            asset_class=asset_class_name,
            as_of_date=as_of_date,
            df_total_return_raw=df_total_return_raw,
            universe_map=universe_map
        )
        tool_schemas.append({
            "type": "function",
            "function": {
                "name": "fetch_historical_stats",
                "description": "Compute the full historical statistics panel for the asset class.",
                "parameters": {"type": "object", "properties": {}, "required": []}
            }
        })

        callable_registry["write_historical_stats_json"] = partial(
            write_historical_stats_json,
            asset_class=asset_class_name,
            artifact_dir=artifact_dir
        )
        tool_schemas.append({
            "type": "function",
            "function": {
                "name": "write_historical_stats_json",
                "description": "Serialise and persist the historical statistics to historical_stats.json.",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "historical_stats": {"type": "object", "description": "Output of fetch_historical_stats."}
                    },
                    "required": ["historical_stats"]
                }
            }
        })

    # ------------------------------------------------------------------
    # Tools applicable to Equity, Real Assets, and Fixed Income
    # ------------------------------------------------------------------
    if category in ["Equity", "Real Assets", "Fixed Income"]:
        callable_registry["fetch_signals"] = partial(
            fetch_signals,
            asset_class=asset_class_name,
            as_of_date=as_of_date,
            df_signals_raw=df_signals_raw,
            df_fundamentals_raw=df_fundamentals_raw,
            universe_map=universe_map
        )
        tool_schemas.append({
            "type": "function",
            "function": {
                "name": "fetch_signals",
                "description": "Retrieve and validate the most recent technical and valuation signals.",
                "parameters": {"type": "object", "properties": {}, "required": []}
            }
        })

        callable_registry["write_signals_json"] = partial(
            write_signals_json,
            asset_class=asset_class_name,
            artifact_dir=artifact_dir
        )
        tool_schemas.append({
            "type": "function",
            "function": {
                "name": "write_signals_json",
                "description": "Serialise and persist the signals dict to signals.json.",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "signals": {"type": "object", "description": "Output of fetch_signals."}
                    },
                    "required": ["signals"]
                }
            }
        })

        callable_registry["write_cma_methods_json"] = partial(
            write_cma_methods_json,
            asset_class=asset_class_name,
            artifact_dir=artifact_dir
        )
        tool_schemas.append({
            "type": "function",
            "function": {
                "name": "write_cma_methods_json",
                "description": "Serialise and persist the CMA method results to cma_methods.json.",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "method_results": {
                            "type": "array",
                            "description": "List of method result dicts from CMA Methods 1-7 or FI builder."
                        }
                    },
                    "required": ["method_results"]
                }
            }
        })

    # ------------------------------------------------------------------
    # Tools applicable to Equity and Real Assets
    # ------------------------------------------------------------------
    if category in ["Equity", "Real Assets"]:
        callable_registry["run_cma_method_1"] = partial(run_cma_method_1, rf=rf)
        tool_schemas.append({
            "type": "function",
            "function": {
                "name": "run_cma_method_1",
                "description": "Compute CMA Method 1: Historical Equity Risk Premium + Risk-Free Rate.",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "historical_stats": {"type": "object", "description": "Output of fetch_historical_stats."}
                    },
                    "required": ["historical_stats"]
                }
            }
        })

        # Inject macro_view directly to prevent LLM tampering
        callable_registry["run_cma_method_2"] = partial(run_cma_method_2, macro_view=macro_view, rf=rf)
        tool_schemas.append({
            "type": "function",
            "function": {
                "name": "run_cma_method_2",
                "description": "Compute CMA Method 2: Regime-Adjusted Equity Risk Premium + Risk-Free Rate.",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "historical_stats": {"type": "object", "description": "Output of fetch_historical_stats."}
                    },
                    "required": ["historical_stats"]
                }
            }
        })

        callable_registry["run_cma_method_6"] = partial(
            run_cma_method_6,
            asset_class=asset_class_name,
            as_of_date=as_of_date,
            df_survey_cma_raw=df_survey_cma_raw
        )
        tool_schemas.append({
            "type": "function",
            "function": {
                "name": "run_cma_method_6",
                "description": "Compute CMA Method 6: Survey/Analyst Consensus Expected Return.",
                "parameters": {"type": "object", "properties": {}, "required": []}
            }
        })

        callable_registry["run_cma_method_7"] = run_cma_method_7
        tool_schemas.append({
            "type": "function",
            "function": {
                "name": "run_cma_method_7",
                "description": "Compute CMA Method 7: Confidence-Weighted Auto-Blend of Methods 1-6.",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "method_results": {
                            "type": "array",
                            "description": "List of method result dicts from Methods 1-6."
                        }
                    },
                    "required": ["method_results"]
                }
            }
        })

    # ------------------------------------------------------------------
    # Tools applicable ONLY to Equity
    # ------------------------------------------------------------------
    if category == "Equity":
        # Inject the mathematically rigorous sigma and w_mkt arrays
        callable_registry["run_cma_method_3"] = partial(
            run_cma_method_3,
            asset_class=asset_class_name,
            sigma=sigma,
            w_mkt=w_mkt
        )
        tool_schemas.append({
            "type": "function",
            "function": {
                "name": "run_cma_method_3",
                "description": "Compute CMA Method 3: Black-Litterman Equilibrium Implied Return.",
                "parameters": {"type": "object", "properties": {}, "required": []}
            }
        })

        callable_registry["run_cma_method_4"] = partial(
            run_cma_method_4,
            asset_class=asset_class_name,
            as_of_date=as_of_date,
            df_fundamentals_raw=df_fundamentals_raw,
            universe_map=universe_map
        )
        tool_schemas.append({
            "type": "function",
            "function": {
                "name": "run_cma_method_4",
                "description": "Compute CMA Method 4: Inverse Gordon / Grinold-Kroner Building-Block Model.",
                "parameters": {"type": "object", "properties": {}, "required": []}
            }
        })

        callable_registry["run_cma_method_5"] = partial(run_cma_method_5, rf=rf)
        tool_schemas.append({
            "type": "function",
            "function": {
                "name": "run_cma_method_5",
                "description": "Compute CMA Method 5: Implied ERP Proxy (CAPE-based Earnings Yield).",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "signals": {"type": "object", "description": "Output of fetch_signals."}
                    },
                    "required": ["signals"]
                }
            }
        })

    # ------------------------------------------------------------------
    # Tools applicable ONLY to Fixed Income
    # ------------------------------------------------------------------
    if category == "Fixed Income":
        # CRITICAL FIX: Inject the new fi_assumptions and duration_map parameters
        callable_registry["run_fi_cma_builder"] = partial(
            run_fi_cma_builder,
            asset_class=asset_class_name,
            as_of_date=as_of_date,
            df_fixed_income_curves_spreads_raw=df_fixed_income_curves_spreads_raw,
            df_total_return_raw=df_total_return_raw,
            universe_map=universe_map,
            fi_assumptions=fi_assumptions,
            duration_map=duration_map
        )
        # The schema exposed to the LLM remains empty, as all parameters are bound via closure
        tool_schemas.append({
            "type": "function",
            "function": {
                "name": "run_fi_cma_builder",
                "description": "Compute the Fixed Income CMA using the building-block decomposition.",
                "parameters": {"type": "object", "properties": {}, "required": []}
            }
        })

        callable_registry["write_cma_json"] = partial(
            write_cma_json,
            asset_class=asset_class_name,
            artifact_dir=artifact_dir,
            as_of_date=as_of_date
        )
        tool_schemas.append({
            "type": "function",
            "function": {
                "name": "write_cma_json",
                "description": "Serialise and persist the final CMA estimate to cma.json.",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "final_estimate": {"type": "number"},
                        "confidence": {"type": "number"},
                        "rationale": {"type": "string"},
                        "method_range": {"type": "object"},
                        "credit_spread_duration": {"type": ["number", "null"]},
                        "sector_concentration": {"type": ["number", "null"]}
                    },
                    "required": ["final_estimate", "confidence", "rationale", "method_range"]
                }
            }
        })

    # ------------------------------------------------------------------
    # Tools applicable ONLY to Cash
    # ------------------------------------------------------------------
    if category == "Cash":
        callable_registry["fetch_cash_yield_proxy"] = partial(
            fetch_cash_yield_proxy,
            as_of_date=as_of_date,
            df_fixed_income_curves_spreads_raw=df_fixed_income_curves_spreads_raw
        )
        tool_schemas.append({
            "type": "function",
            "function": {
                "name": "fetch_cash_yield_proxy",
                "description": "Retrieve the current cash yield proxy.",
                "parameters": {"type": "object", "properties": {}, "required": []}
            }
        })

        callable_registry["write_cma_json"] = partial(
            write_cma_json,
            asset_class=asset_class_name,
            artifact_dir=artifact_dir,
            as_of_date=as_of_date,
            method_weights={}
        )
        tool_schemas.append({
            "type": "function",
            "function": {
                "name": "write_cma_json",
                "description": "Serialise and persist the final CMA estimate to cma.json.",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "final_estimate": {"type": "number"},
                        "confidence": {"type": "number"},
                        "rationale": {"type": "string"},
                        "method_range": {"type": "object"}
                    },
                    "required": ["final_estimate", "confidence", "rationale", "method_range"]
                }
            }
        })

        callable_registry["write_analysis_md"] = partial(
            write_analysis_md,
            artifact_dir=artifact_dir,
            as_of_date=as_of_date
        )
        tool_schemas.append({
            "type": "function",
            "function": {
                "name": "write_analysis_md",
                "description": "Format and persist the analysis narrative to analysis.md.",
                "parameters": {
                    "type": "object",
                    "properties": {
                        "regime": {"type": "string"},
                        "rationale": {"type": "string"}
                    },
                    "required": ["regime", "rationale"]
                }
            }
        })

    logger.debug(f"Successfully registered tools for AC Agent {asset_class_name}.")
    return callable_registry, tool_schemas

# -------------------------------------------------------------------------------------------------------------------------------
# Task 18, Step 4: Execute independent parallel ReAct loops and bind output contracts
# -------------------------------------------------------------------------------------------------------------------------------

def execute_ac_react_loop(
    asset_class_name: str,
    client: Any, # OpenAI Client
    llm_config: AgentLLMConfig,
    compiled_prompt: str,
    callable_registry: Dict[str, Callable],
    tool_schemas: List[Dict[str, Any]],
    termination_tools: List[str],
    max_iterations: int = 30
) -> Dict[str, str]:
    """
    Executes the ReAct loop for a single AC Agent. Manages state accumulation,
    tool dispatch, and strict termination conditions.

    Parameters
    ----------
    asset_class_name : str
        The name of the asset class.
    client : Any
        The authenticated OpenAI client.
    llm_config : AgentLLMConfig
        The strictly typed configuration for the agent.
    compiled_prompt : str
        The compiled developer prompt.
    callable_registry : Dict[str, Callable]
        The dictionary mapping tool names to Python functions.
    tool_schemas : List[Dict[str, Any]]
        The JSON Schema array of available tools.
    termination_tools : List[str]
        The list of tools that must be called before termination.
    max_iterations : int, optional
        The hard iteration ceiling. Defaults to 30.

    Returns
    -------
    Dict[str, str]
        A dictionary mapping artifact names to their written file paths.

    Raises
    ------
    RuntimeError
        If the iteration ceiling is hit without clean termination.
    """
    # Initialize the conversation history
    messages: List[Dict[str, str]] = [
        {"role": "developer", "content": compiled_prompt},
        {"role": "user", "content": f"Begin CMA analysis for {asset_class_name}. Execute the workflow strictly in order."}
    ]

    # State trackers for termination conditions
    tools_called: set = set()
    artifact_paths: Dict[str, str] = {}

    # Local state for accumulating method results (specifically for method 7)
    local_state: Dict[str, Any] = {"method_results": []}

    iteration: int = 0
    while iteration < max_iterations:
        iteration += 1
        logger.debug(f"AC Agent ({asset_class_name}) ReAct Loop Iteration {iteration}/{max_iterations}")

        # Invoke the LLM
        response: ParsedLLMResponse = invoke_and_extract_agent_response(
            client=client,
            config=llm_config,
            messages=messages,
            tools=tool_schemas
        )

        # Append the assistant's response to the history
        assistant_msg: Dict[str, Any] = {"role": "assistant"}
        if response.content:
            assistant_msg["content"] = response.content
        if response.tool_calls:
            assistant_msg["tool_calls"] = response.tool_calls
        messages.append(assistant_msg)

        # Check if tools were called
        if response.tool_calls:
            for tool_call in response.tool_calls:
                tool_name: str = tool_call.function.name

                # Parse the JSON arguments provided by the LLM with strict error handling
                try:
                    tool_args: Dict[str, Any] = json.loads(tool_call.function.arguments)
                except json.JSONDecodeError as exc:
                    # Catch JSON errors and feed them back to the LLM to force correction
                    error_msg: str = f"JSONDecodeError: Failed to parse arguments for tool {tool_name}. Details: {exc}"
                    logger.error(f"AC Agent ({asset_class_name}) - {error_msg}")
                    messages.append({
                        "role": "tool",
                        "tool_call_id": tool_call.id,
                        "name": tool_name,
                        "content": error_msg
                    })
                    continue # Skip execution and let the LLM retry

                if tool_name in callable_registry:
                    try:
                        # Inject accumulated method results for run_cma_method_7
                        if tool_name == "run_cma_method_7":
                            tool_args["method_results"] = local_state["method_results"]
                        elif tool_name == "write_cma_methods_json":
                            # Override LLM hallucinated results with our accumulated state
                            tool_args["method_results"] = local_state["method_results"]

                        # Execute the deterministic Python script
                        result: Any = callable_registry[tool_name](**tool_args)

                        # Accumulate method results
                        if tool_name.startswith("run_cma_method_") and tool_name != "run_cma_method_7":
                            local_state["method_results"].append(result)
                        elif tool_name == "run_fi_cma_builder":
                            # FI builder acts as the primary method
                            result["method_id"] = 1
                            local_state["method_results"].append(result)

                        # Track termination conditions and artifact paths
                        if tool_name in termination_tools:
                            tools_called.add(tool_name)
                            artifact_paths[tool_name] = str(result)

                        # Serialize the result for the LLM
                        result_str: str = json.dumps(result)
                    except Exception as exc:
                        # Catch execution errors and feed them back to the LLM
                        result_str = f"Error executing {tool_name}: {str(exc)}"
                        logger.error(f"AC Agent ({asset_class_name}) - {result_str}")
                else:
                    result_str = f"Error: Tool {tool_name} not found in registry."

                messages.append({
                    "role": "tool",
                    "tool_call_id": tool_call.id,
                    "name": tool_name,
                    "content": result_str
                })

        # Check termination conditions
        if all(tool in tools_called for tool in termination_tools):
            logger.info(f"AC Agent ({asset_class_name}) ReAct loop terminated cleanly.")
            return artifact_paths

        # Prompt to continue if stalled
        if not response.tool_calls and not all(tool in tools_called for tool in termination_tools):
            missing_tools = [t for t in termination_tools if t not in tools_called]
            messages.append({
                "role": "user",
                "content": f"You have not completed the workflow. You must call {missing_tools} before terminating."
            })

    raise RuntimeError(f"AC Agent ({asset_class_name}) ReAct loop hit iteration ceiling ({max_iterations}).")

# -------------------------------------------------------------------------------------------------------------------------------
# Task 18, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_18_ac_agents(
    df_total_return_raw: pd.DataFrame,
    df_signals_raw: pd.DataFrame,
    df_fundamentals_raw: pd.DataFrame,
    df_survey_cma_raw: pd.DataFrame,
    df_fixed_income_curves_spreads_raw: pd.DataFrame,
    df_benchmark_factors_raw: pd.DataFrame,
    macro_view_path: str,
    sigma: np.ndarray,
    config: Dict[str, Any],
    run_output_dir: Path,
    client: Any, # OpenAI Client
    validate_artifact_fn: Callable[[Path, str], bool]
) -> Dict[str, Dict[str, str]]:
    """
    Orchestrates the AC Agent packages, executing parallel ReAct loops for all 18
    asset classes and applying hard schema validation gates on the outputs.

    CRITICAL FIX: Extracts fi_assumptions and duration_map from the configuration
    and passes them to register_ac_agent_tools, ensuring the FI CMA builder
    operates under strict, configurable mathematical assumptions.

    Parameters
    ----------
    df_total_return_raw : pd.DataFrame
        The cleansed total return dataframe.
    df_signals_raw : pd.DataFrame
        The cleansed signals dataframe.
    df_fundamentals_raw : pd.DataFrame
        The cleansed fundamentals dataframe.
    df_survey_cma_raw : pd.DataFrame
        The cleansed survey CMA dataframe.
    df_fixed_income_curves_spreads_raw : pd.DataFrame
        The cleansed fixed income curves dataframe.
    df_benchmark_factors_raw : pd.DataFrame
        Required to extract the point-in-time risk-free rate.
    macro_view_path : str
        Path to the macro-view.json artifact.
    sigma : np.ndarray
        The 18x18 Ledoit-Wolf covariance matrix required for Black-Litterman.
    config : Dict[str, Any]
        The validated study configuration dictionary.
    run_output_dir : Path
        The root directory for the current run's artifacts.
    client : Any
        The authenticated OpenAI client.
    validate_artifact_fn : Callable[[Path, str], bool]
        The bound validation function from Task 16.

    Returns
    -------
    Dict[str, Dict[str, str]]
        A dictionary mapping asset class names to their generated artifact paths.

    Raises
    ------
    RuntimeError
        If any AC agent fails execution or validation (fail-closed).
    KeyError
        If required configuration keys or universe mappings are missing.
    """
    logger.info("Starting Task 18: Construct AC Agent packages (equity, fixed income, real assets, cash).")

    # Step 1: Extract configuration parameters and generate IPS Markdown
    try:
        ips_config: Dict[str, Any] = config["IPS_GOVERNANCE"]
        ips_path: str = generate_ips_markdown(ips_config, run_output_dir)

        as_of_date: str = config["AS_OF_DATE"]
        universe_map: Dict[str, Dict[str, Any]] = config["REGISTRIES"]["universe_map"]

        asset_classes_dict: Dict[str, List[str]] = config["IPS_GOVERNANCE"]["ASSET_CLASSES"]
        canonical_order: List[str] = []
        for category in ["equity", "fixed_income", "real_assets", "cash"]:
            canonical_order.extend(asset_classes_dict.get(category, []))

        # CRITICAL FIX: Extract new FI parameters from the methodology configuration
        # Fail-closed if these fiduciary parameters are missing
        fi_assumptions: Dict[str, Dict[str, float]] = config["METHODOLOGY_PARAMS"]["FI_ASSUMPTIONS"]
        duration_map: Dict[str, float] = config["METHODOLOGY_PARAMS"]["DURATION_MAP"]

    except KeyError as exc:
        raise KeyError(f"Configuration is missing required keys for Task 18: {exc}") from exc

    # Extract the point-in-time risk-free rate
    rf: float = extract_current_risk_free_rate(df_benchmark_factors_raw, as_of_date)

    # Extract the market-cap weights for Black-Litterman using robust imputation logic
    w_mkt: np.ndarray = extract_market_cap_weights(
        df_fundamentals_raw=df_fundamentals_raw,
        as_of_date=as_of_date,
        canonical_order=canonical_order,
        universe_map=universe_map
    )

    # Load macro view for injection into the tools
    try:
        with open(macro_view_path, "r", encoding="utf-8") as f:
            macro_view: Dict[str, Any] = json.load(f)
    except Exception as exc:
        raise RuntimeError(f"Failed to load macro view from {macro_view_path}: {exc}") from exc

    # Extract LLM and Agent configurations
    prompt_templates: Dict[str, Any] = config["PROMPT_TEMPLATES"]
    llm_policy_dict: Dict[str, Any] = config["INFRASTRUCTURE"]["LLM_INVOCATION_POLICY"]
    effort_schedules: Dict[str, Any] = config["INFRASTRUCTURE"]["REASONING_EFFORT_SCHEDULE"]
    react_params: Dict[str, Any] = config["INFRASTRUCTURE"]["REACT_LOOP_PARAMS"]
    ac_configs: Dict[str, Any] = config["REGISTRIES"]["AC_AGENTS"]["asset_class_configs"]

    # Define the specific artifact directory for AC Agents
    ac_artifact_dir: Path = run_output_dir / "ac_agents"

    # Inner function to process a single asset class within a thread
    def process_asset_class(ac_name: str, ac_config: Dict[str, Any]) -> Tuple[str, Dict[str, str]]:
        logger.info(f"Initializing AC Agent for {ac_name}...")

        # Determine LLM config based on category
        policy_key: str = ac_config["react_params_key"]
        effort_schedule = effort_schedules[policy_key]
        llm_config = AgentLLMConfig(
            model_name=llm_policy_dict["model"],
            reasoning_effort=effort_schedule["effort"],
            max_completion_tokens=effort_schedule["max_output_tokens"]
        )

        termination_tools: List[str] = react_params[policy_key]["termination_tools"]
        max_iterations: int = react_params[policy_key]["max_iterations"]

        # Step 2: Compile the prompt
        compiled_prompt: str = compile_ac_agent_prompt(
            asset_class_config=ac_config,
            prompt_templates=prompt_templates,
            macro_view_path=macro_view_path,
            as_of_date=as_of_date,
            rf=rf,
            ips_path=ips_path
        )

        # Step 3: Register tools
        # CRITICAL FIX: Pass the new FI parameters to the registry
        callable_registry, tool_schemas = register_ac_agent_tools(
            asset_class_name=ac_name,
            category=ac_config["type"].replace("_", " ").title(),
            df_total_return_raw=df_total_return_raw,
            df_signals_raw=df_signals_raw,
            df_fundamentals_raw=df_fundamentals_raw,
            df_survey_cma_raw=df_survey_cma_raw,
            df_fixed_income_curves_spreads_raw=df_fixed_income_curves_spreads_raw,
            universe_map=universe_map,
            macro_view=macro_view,
            rf=rf,
            sigma=sigma,
            w_mkt=w_mkt,
            artifact_dir=ac_artifact_dir,
            as_of_date=as_of_date,
            fi_assumptions=fi_assumptions,
            duration_map=duration_map
        )

        # Step 4: Execute ReAct loop
        artifact_paths: Dict[str, str] = execute_ac_react_loop(
            asset_class_name=ac_name,
            client=client,
            llm_config=llm_config,
            compiled_prompt=compiled_prompt,
            callable_registry=callable_registry,
            tool_schemas=tool_schemas,
            termination_tools=termination_tools,
            max_iterations=max_iterations
        )

        # Apply hard schema validation gates
        for tool_name, path_str in artifact_paths.items():
            path = Path(path_str)
            if tool_name == "write_cma_methods_json":
                validate_artifact_fn(path, "cma_methods")
            elif tool_name == "write_cma_json":
                validate_artifact_fn(path, "cma")
            elif tool_name == "write_signals_json":
                validate_artifact_fn(path, "signals")
            elif tool_name == "write_historical_stats_json":
                validate_artifact_fn(path, "historical_stats")

        return ac_name, artifact_paths

    # Execute all 18 AC agents in parallel with strict exception handling
    all_artifacts: Dict[str, Dict[str, str]] = {}
    max_workers: int = config["INFRASTRUCTURE"]["AUTOGEN_TOPOLOGY"]["parallel_dispatch"].get("max_workers_ac_agents", 8)

    with concurrent.futures.ThreadPoolExecutor(max_workers=max_workers) as executor:
        # Submit all tasks
        future_to_ac = {
            executor.submit(process_asset_class, ac_name, ac_config): ac_name
            for ac_name, ac_config in ac_configs.items()
        }

        # Collect results as they complete, enforcing fail-closed semantics
        for future in concurrent.futures.as_completed(future_to_ac):
            ac_name = future_to_ac[future]
            try:
                # future.result() will re-raise any exception caught within the thread
                ac_name_result, artifact_paths = future.result()
                all_artifacts[ac_name_result] = artifact_paths
                logger.info(f"AC Agent for {ac_name} completed successfully.")
            except Exception as exc:
                logger.error(f"AC Agent for {ac_name} generated a fatal exception: {exc}")
                # Fail-closed: if one agent fails, the pipeline must halt immediately
                raise RuntimeError(f"Parallel execution failed for AC Agent {ac_name}: {exc}") from exc

    # Final verification: ensure exactly 18 agents completed
    if len(all_artifacts) != 18:
        raise RuntimeError(f"Expected 18 AC Agent completions, but only {len(all_artifacts)} succeeded.")

    logger.info("Task 18 completed successfully. All 18 AC Agent packages executed and validated.")

    return all_artifacts
````

````python
# Task 19 — Construct PC Agents, CRO Agent, Strategy Review controller, CIO Agent

# ==============================================================================
# Task 19: Construct PC Agents, CRO Agent, Strategy Review controller, CIO Agent
# ==============================================================================

# ==============================================================================
# Custom Exceptions and Types
# ==============================================================================

class SchemaValidationError(Exception):
    """
    Raised when a JSON artifact fails schema validation at an inter-phase
    transition gate.

    This exception class is used throughout the pipeline to implement the
    fail-closed architecture: any caller catching it must halt the pipeline
    immediately. It inherits from Exception (not RuntimeError) so that it can
    be distinguished from unexpected runtime failures in broad except clauses,
    allowing the orchestrator to differentiate between data errors and strict
    machine-to-machine contract violations.
    """
    pass


class MASTopology(NamedTuple):
    """
    Named container for the fully instantiated AutoGen Multi-Agent System topology.

    This data structure encapsulates the entire state of the multi-agent Directed
    Acyclic Graph (DAG) required for the Strategy Review pipeline. It ensures that
    all agent instances, communication channels, and state variables are securely
    passed between the orchestration phases.

    CRITICAL REMEDIATION: Added `pipeline_state` dictionary to allow the master
    orchestrator to pass runtime variables (like top5_slugs) to downstream agents
    (like the CIO) that are instantiated before the variables exist, thereby
    resolving temporal paradoxes in the DAG.
    """
    # Nested dictionary mapping phase labels (e.g., 'Phase_A') to agent slugs and their ConversableAgent instances
    agents: Dict[str, Dict[str, autogen.ConversableAgent]]

    # The AutoGen GroupChat object containing all topology participants for centralized message routing
    group_chat: autogen.GroupChat

    # The StrategyReviewController managing the deterministic Finite State Machine (FSM)
    group_chat_manager: autogen.GroupChatManager

    # The live FSM instance bound to the GroupChatManager to track phase transitions
    fsm_controller: Any

    # Factory callables keyed by agent slug to enable race-condition-free parallel dispatch
    agent_factories: Dict[str, Any]

    # The raw CRO system prompt template stored for dynamic per-candidate formatting at runtime
    cro_prompt_template: str

    # The full Portfolio Construction (PC) method registry defining the 21 candidate algorithms
    pc_registry: Dict[str, Any]

    # Mapping from method slug to the 18-element expected-return vector (Capital Market Assumptions)
    mu_dict: Dict[str, np.ndarray]

    # Mutable state dictionary for resolving temporal paradoxes across sequential DAG phases
    pipeline_state: Dict[str, Any]

# ==============================================================================
# Helper Functions for Topology Instantiation
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Helper: Termination Message Factory
# -------------------------------------------------------------------------------------------------------------------------------

def _make_termination_fn(
    terminal_tools: List[str],
    termination_signal: str = "TASK_COMPLETE"
) -> Callable[[Dict[str, Any]], bool]:
    """
    Factory that produces an `is_termination_msg` callable compatible with
    autogen.ConversableAgent's termination parameter.

    This function enforces strict operational boundaries by terminating an agent's
    reasoning loop only when specific, mathematically required tools have been
    invoked, or when an explicit completion signal is detected in the text.

    Parameters
    ----------
    terminal_tools : List[str]
        A list of string identifiers for the tools that, when called, signal the
        successful completion of the agent's task (e.g., 'write_cma_json').
    termination_signal : str, optional
        A specific string token that, if present in the LLM's text output, triggers
        termination. Defaults to "TASK_COMPLETE".

    Returns
    -------
    Callable[[Dict[str, Any]], bool]
        A deterministic, zero-side-effect predicate function that evaluates a
        message dictionary and returns True if termination conditions are met.
    """
    # Convert the list of terminal tools to a frozenset to guarantee O(1) membership
    # testing and ensure the target set remains immutable during execution.
    _terminal_set: frozenset = frozenset(terminal_tools)

    # Capture the termination signal as a local variable inside the closure for fast access.
    _signal: str = termination_signal

    def _termination_fn(message: Dict[str, Any]) -> bool:
        """
        Evaluates a single message dictionary to determine if the conversation
        should terminate based on content or tool calls.
        """
        # Extract the message content; default to an empty string if the key is absent to prevent TypeErrors.
        content: str = message.get("content") or ""

        # Condition (a): Check if the explicit termination signal is present in the textual content.
        if _signal in content:
            # Return True to immediately halt the agent's reasoning loop.
            return True

        # Extract the tool_calls list; default to an empty list if the key is absent.
        tool_calls: List[Dict[str, Any]] = message.get("tool_calls") or []

        # Condition (b): Iterate through all tool calls requested by the LLM in this message.
        for tc in tool_calls:
            # Safely extract the invoked function name from the nested tool call dictionary.
            fn_name: str = tc.get("function", {}).get("name", "")

            # If the invoked function name exists in the immutable set of terminal tools, signal completion.
            if fn_name in _terminal_set:
                # Return True to halt the loop, as the required output contract has been fulfilled.
                return True

        # If neither the text signal nor a terminal tool call is detected, the conversation must continue.
        return False

    # Return the constructed termination function closure to be bound to the ConversableAgent.
    return _termination_fn


# -------------------------------------------------------------------------------------------------------------------------------
# Helper: LLM CONFIG Builder
# -------------------------------------------------------------------------------------------------------------------------------

def _create_llm_config(
    policy_key: str,
    config: Dict[str, Any],
    openai_api_key: str,
    force_tool_choice: Optional[Dict[str, Any]] = None
) -> Dict[str, Any]:
    """
    Constructs the AutoGen `llm_config` dictionary from the frozen
    REASONING_EFFORT_SCHEDULE and LLM_INVOCATION_POLICY in STUDY_CONFIG.

    This function maps the conceptual effort levels defined in the institutional
    configuration to the strict API parameters required by GPT-5.2, ensuring
    deterministic compute allocation across the multi-agent system.

    Parameters
    ----------
    policy_key : str
        The specific key into INFRASTRUCTURE.REASONING_EFFORT_SCHEDULE for the target agent.
    config : Dict[str, Any]
        The validated, immutable STUDY_CONFIG dictionary.
    openai_api_key : str
        The OpenAI API key, sourced securely from the environment.
    force_tool_choice : Optional[Dict[str, Any]], optional
        An AutoGen-level tool_choice override injected at the llm_config level to
        force structured outputs (e.g., for voting agents). Defaults to None.

    Returns
    -------
    Dict[str, Any]
        A compliant llm_config dictionary ready for injection into an autogen.ConversableAgent.

    Raises
    ------
    KeyError
        If the requested policy_key is absent from the REASONING_EFFORT_SCHEDULE.
    """
    # Extract the frozen top-level LLM invocation policy (containing model name and API surface).
    llm_policy: Dict[str, Any] = config["INFRASTRUCTURE"]["LLM_INVOCATION_POLICY"]

    # Extract the full effort schedule to validate the existence of the requested policy key.
    effort_schedule: Dict[str, Any] = config["INFRASTRUCTURE"]["REASONING_EFFORT_SCHEDULE"]

    # Guard: Fail explicitly and immediately if the policy_key is not registered in the schedule.
    if policy_key not in effort_schedule:
        # Raise a KeyError to halt instantiation and flag the configuration defect.
        raise KeyError(f"_create_llm_config: policy_key '{policy_key}' not found in REASONING_EFFORT_SCHEDULE.")

    # Extract this specific agent's effort tier and output token budget from the schedule.
    agent_effort: Dict[str, Any] = effort_schedule[policy_key]

    # Build the per-model entry for the AutoGen config_list using strict GPT-5.2 Responses API parameters.
    model_entry: Dict[str, Any] = {
        # Assign the exact model identifier (e.g., 'gpt-5.2').
        "model": llm_policy["model"],
        # Inject the securely sourced API key.
        "api_key": openai_api_key,
        # Define the API surface to utilize (e.g., 'responses').
        "api_surface": llm_policy["api_surface"],
        # Map the reasoning effort tier (e.g., 'high', 'xhigh') to the API payload.
        "reasoning": {"effort": agent_effort["effort"]},
        # Set the absolute ceiling for generated tokens to prevent runaway generation.
        "max_output_tokens": agent_effort["max_output_tokens"],
    }

    # Construct the top-level llm_config dictionary accepted by autogen.ConversableAgent.
    llm_cfg: Dict[str, Any] = {
        # Provide the single-entry config_list.
        "config_list": [model_entry],
        # Explicitly disable result caching to ensure each invocation produces a fresh, deterministic reasoning trace.
        "cache_seed": None,
    }

    # Inject the forced tool_choice if provided (used exclusively to guarantee structured JSON from voting agents).
    if force_tool_choice is not None:
        # Append the tool_choice directive to the configuration dictionary.
        llm_cfg["tool_choice"] = force_tool_choice

    # Return the fully constructed and validated configuration dictionary.
    return llm_cfg

# -------------------------------------------------------------------------------------------------------------------------------
# Finite State Machine: Strategy Review Controller
# -------------------------------------------------------------------------------------------------------------------------------

class _StrategyReviewFSMController:
    """
    Implements the deterministic Finite State Machine (FSM) for the
    StrategyReviewController (GroupChatManager).

    By overriding the default LLM-based speaker selection with this deterministic
    FSM, we guarantee that the multi-agent deliberation protocol executes in the
    exact sequence mandated by the architecture (Phases A through G). This prevents
    agents from speaking out of turn or bypassing critical fail-closed validation gates.
    """
    # Define the canonical phase sequence as an immutable tuple to prevent accidental mutation during runtime.
    PHASE_SEQUENCE: Tuple[str, ...] = ("A", "B", "C", "D", "E", "F", "G")

    def __init__(
        self,
        phase_speaker_map: Dict[str, List[str]],
        artifact_dir: Path,
        validate_artifact_fn: Callable[[Path, str], bool],
        all_agents_by_name: Dict[str, autogen.ConversableAgent],
        config: Dict[str, Any],
    ) -> None:
        """
        Initializes the FSM Controller with the required state mappings and validation tools.

        Parameters
        ----------
        phase_speaker_map : Dict[str, List[str]]
            A dictionary mapping each phase label to the list of agent names expected to speak.
        artifact_dir : Path
            The root artifact directory for the current run, used for validation checks.
        validate_artifact_fn : Callable[[Path, str], bool]
            The bound schema-validation callable used to enforce fail-closed gates.
        all_agents_by_name : Dict[str, autogen.ConversableAgent]
            A dictionary mapping agent names to their instantiated ConversableAgent objects.
        config : Dict[str, Any]
            The validated STUDY_CONFIG dictionary.
        """
        # Track the integer index of the currently active phase, starting at 0.
        self._phase_idx: int = 0

        # Set the current phase label derived from the canonical sequence (starts at 'A').
        self._current_phase: str = self.PHASE_SEQUENCE[0]

        # Initialize an empty set to track agent names that have completed their turn in the current phase.
        self._completed_this_phase: Set[str] = set()

        # Store the map of phase labels to expected speaker names.
        self._phase_speaker_map: Dict[str, List[str]] = phase_speaker_map

        # Store the artifact directory path for inter-phase validation calls.
        self._artifact_dir: Path = artifact_dir

        # Store the bound validation callable to enforce fail-closed schema gating.
        self._validate_fn: Callable = validate_artifact_fn

        # Store the dictionary of agent references for dynamic speaker selection.
        self._agents: Dict[str, autogen.ConversableAgent] = all_agents_by_name

        # Preserve the configuration reference for potential threshold access.
        self._config: Dict[str, Any] = config

        # Initialize the terminal flag; this will be set to True once Phase G is completely validated.
        self._pipeline_complete: bool = False

    def __call__(
        self,
        last_speaker: autogen.ConversableAgent,
        groupchat: autogen.GroupChat,
    ) -> Optional[autogen.ConversableAgent]:
        """
        Selects the next agent to speak. Called by the GroupChatManager after
        each completed agent turn.

        Parameters
        ----------
        last_speaker : autogen.ConversableAgent
            The agent that just completed its turn.
        groupchat : autogen.GroupChat
            The GroupChat object managing the conversation.

        Returns
        -------
        Optional[autogen.ConversableAgent]
            The next agent to speak, or None to terminate the GroupChat when the
            pipeline is complete or a validation error occurs.
        """
        # If the pipeline is already marked as complete, return None to signal GroupChat termination.
        if self._pipeline_complete:
            return None

        # If a speaker just finished, record their name in the completion set for the current phase.
        if last_speaker is not None:
            self._completed_this_phase.add(last_speaker.name)

        # Retrieve the list of agents expected to speak in the currently active phase.
        expected_speakers: List[str] = self._phase_speaker_map.get(self._current_phase, [])

        # Check whether all expected speakers have successfully completed their turns.
        phase_done: bool = all(name in self._completed_this_phase for name in expected_speakers)

        # If the phase is complete, execute validation and transition logic.
        if phase_done:
            # Execute the fail-closed schema validation gate for this specific phase.
            try:
                self._validate_phase_artifacts(self._current_phase)
            except Exception as exc:
                # Log the critical failure and return None to immediately halt the pipeline (fail-closed).
                logger.critical("StrategyReviewFSM: Phase %s validation failed. %s", self._current_phase, exc)
                return None

            # Advance the internal phase index to point to the next phase.
            self._phase_idx += 1

            # Check if the advanced index exceeds the length of the canonical phase sequence.
            if self._phase_idx >= len(self.PHASE_SEQUENCE):
                # All seven phases are complete; set the terminal flag to True.
                self._pipeline_complete = True
                # Log the successful completion of the entire FSM pipeline.
                logger.info("StrategyReviewFSM: All phases complete.")
                # Return None to gracefully terminate the GroupChat.
                return None

            # Update the current phase label to the new phase.
            self._current_phase = self.PHASE_SEQUENCE[self._phase_idx]

            # Reset the completed-speaker tracker set for the new phase.
            self._completed_this_phase = set()

            # Log the successful transition to the next phase.
            logger.info("StrategyReviewFSM: Advanced to Phase %s.", self._current_phase)

        # Identify which expected speakers in the current phase have not yet spoken.
        remaining: List[str] = [
            name for name in self._phase_speaker_map.get(self._current_phase, [])
            if name not in self._completed_this_phase
        ]

        # If there are no remaining speakers (e.g., due to a configuration error), return None to avoid infinite loops.
        if not remaining:
            return None

        # Select the first remaining agent from the list to be the next speaker.
        next_name: str = remaining[0]

        # Return the actual ConversableAgent object corresponding to the selected name.
        return self._agents.get(next_name)

    def _validate_phase_artifacts(self, phase: str) -> None:
        """
        Executes the schema validation gate for the completed phase.

        This acts as a belt-and-suspenders check for phases that require
        strict artifact validation before downstream agents can proceed.

        Parameters
        ----------
        phase : str
            The single-character label of the phase that just completed (e.g., 'B', 'G').

        Raises
        ------
        SchemaValidationError
            If the artifact fails validation, which is caught by __call__ to halt the pipeline.
        """
        # If Phase B (Adversarial Diversifier) just completed, validate its output.
        if phase == "B":
            # Construct the absolute path to the adversarial agent's pc_weights.json artifact.
            path: Path = self._artifact_dir / "pc_agents" / "adversarial_diversifier" / "pc_weights.json"

            # Invoke the bound validation function to enforce the schema contract.
            self._validate_fn(path, "pc_weights")

        # If Phase G (CIO Agent) just completed, validate its final output.
        elif phase == "G":
            # Construct the absolute path to the CIO agent's final_weights.json artifact.
            path = self._artifact_dir / "cio" / "final_weights.json"

            # Invoke the bound validation function to enforce the schema contract.
            self._validate_fn(path, "final_weights")


# -------------------------------------------------------------------------------------------------------------------------------
# Task 19, Step 1 — Instantiate MAS Topology
# -------------------------------------------------------------------------------------------------------------------------------

def instantiate_mas_topology(
    config:            Dict[str, Any],
    sigma:             np.ndarray,
    mu_dict:           Dict[str, np.ndarray],
    rf:                float,
    benchmark_weights: np.ndarray,
    returns_matrix:    np.ndarray,
    returns_index:     pd.DatetimeIndex,
    factors_df:        pd.DataFrame,
    macro_view_path:   Path,
    cma_dir:           Path,
    artifact_dir:      Path,
    openai_api_key:    str,
    validate_artifact_fn: Callable[[Path, str], bool],
) -> MASTopology:
    """
    Instantiates all AutoGen ConversableAgent objects and the GroupChat /
    GroupChatManager topology required for the full Strategy Review pipeline.

    CRITICAL REMEDIATION: Implements strict Dependency Injection via functools.partial for every
    single tool across all 7 phases. Resolves the CRO Data Starvation defect
    by reading weights from disk inside the factory. Resolves the CIO Temporal
    Paradox by introducing a mutable pipeline_state dictionary.

    Parameters
    ----------
    config : Dict[str, Any]
        The validated STUDY_CONFIG dictionary.
    sigma : np.ndarray
        18x18 annualized shrinkage covariance matrix.
    mu_dict : Dict[str, np.ndarray]
        Mapping from method slug to 18-element expected-return vector.
    rf : float
        Annualized risk-free rate.
    benchmark_weights : np.ndarray
        18-element benchmark weight vector.
    returns_matrix : np.ndarray
        T x 18 monthly returns matrix.
    returns_index : pd.DatetimeIndex
        The temporal index corresponding to the returns_matrix.
    factors_df : pd.DataFrame
        Monthly Fama-French factor returns dataframe.
    macro_view_path : Path
        Filesystem path to the validated macro-view.json artifact.
    cma_dir : Path
        Directory containing all per-asset cma.json artifacts.
    artifact_dir : Path
        Root directory for all run artifacts.
    openai_api_key : str
        OpenAI API key sourced from the environment.
    validate_artifact_fn : Callable[[Path, str], bool]
        Bound schema-validation callable.

    Returns
    -------
    MASTopology
        Named tuple containing the fully instantiated topology.

    Raises
    ------
    ValueError
        If mu_dict is missing a required entry for a uses_cmas=True method.
    FileNotFoundError
        If macro_view_path or cma_dir does not exist.
    """
    logger.info("Task 19 Step 1: Instantiating MAS topology (Fully Remediated).")

    # ------------------------------------------------------------------
    # Pre-flight Checks and Data Loading
    # ------------------------------------------------------------------
    # Verify macro_view_path exists before loading
    if not macro_view_path.exists():
        raise FileNotFoundError(f"macro_view_path not found: {macro_view_path}")

    # Load the macro view JSON once; used in prompt formatting and tool injection
    with open(macro_view_path, "r", encoding="utf-8") as _fh:
        macro_view_data: Dict[str, Any] = json.load(_fh)

    # Verify cma_dir exists before attempting CMA data loading
    if not cma_dir.is_dir():
        raise FileNotFoundError(f"cma_dir not found: {cma_dir}")

    # Extract canonical order for CIO tools
    asset_classes_dict: Dict[str, List[str]] = config["IPS_GOVERNANCE"]["ASSET_CLASSES"]
    canonical_order: List[str] = []
    for category in ["equity", "fixed_income", "real_assets", "cash"]:
        canonical_order.extend(asset_classes_dict.get(category, []))

    # Load all per-asset CMA estimates from disk for CIO tool injection
    cma_set: Dict[str, float] = {}
    for asset_class in asset_classes_dict.get("equity", []):
        slug: str = asset_class.lower().replace(" ", "_")
        cma_path: Path = cma_dir / slug / "cma.json"
        if cma_path.exists():
            with open(cma_path, "r", encoding="utf-8") as _cf:
                _cma_data: Dict[str, Any] = json.load(_cf)
            cma_set[asset_class] = float(_cma_data["final_estimate"])
        else:
            logger.warning("CMA artifact missing for '%s': %s", asset_class, cma_path)

    # Extract key configuration sub-dicts
    prompt_templates: Dict[str, Any] = config["PROMPT_TEMPLATES"]
    ips_constraints:  Dict[str, Any] = config["IPS_GOVERNANCE"]["IMPLEMENTATION_CONSTRAINTS"]
    pc_registry:      Dict[str, Any] = config["REGISTRIES"]["PC_AGENTS"]["method_configs"]
    react_params:     Dict[str, Any] = config["INFRASTRUCTURE"]["REACT_LOOP_PARAMS"]
    allowed_domains:  List[str] = config["INFRASTRUCTURE"]["LOOKAHEAD_CONTROLS"]["web_search_allowed_domains_baseline"]
    as_of_date:       str = config["AS_OF_DATE"]

    # ------------------------------------------------------------------
    # CRITICAL REMEDIATION: Dynamic State Dictionary
    # ------------------------------------------------------------------
    # This dictionary resolves the temporal paradox by allowing upstream phases
    # (like Phase F) to populate variables (like top5_slugs) that downstream
    # agents (like the CIO) will access at runtime.
    pipeline_state: Dict[str, Any] = {}

    # Containers for agents and factories
    agents: Dict[str, Dict[str, autogen.ConversableAgent]] = {
        "Phase_A": {}, "Phase_B": {}, "Phase_C": {}, "Phase_D": {},
        "Phase_E": {}, "Phase_F": {}, "Phase_G": {}
    }
    agent_factories: Dict[str, Any] = {}

    # ==================================================================
    # PHASE A: PC GENERIC AGENTS AND PC RESEARCHER
    # ==================================================================
    logger.info("Instantiating Phase A: PC Generic and Researcher agents.")

    def _register_researcher_tools(agent: autogen.ConversableAgent) -> None:
        """Binds tools for the PC Researcher Agent."""
        autogen.agentchat.register_function(
            partial(tool_literature_search, as_of_date=as_of_date, allowed_domains=allowed_domains),
            caller=agent, executor=agent, name="tool_literature_search",
            description="Search academic literature for novel portfolio construction methods."
        )
        autogen.agentchat.register_function(
            validate_method_spec,
            caller=agent, executor=agent, name="validate_method_spec",
            description="Validate that a proposed method specification is IPS-compatible."
        )
        autogen.agentchat.register_function(
            partial(write_proposed_method_md, artifact_dir=artifact_dir / "pc_agents" / "researcher"),
            caller=agent, executor=agent, name="write_proposed_method_md",
            description="Persist the proposed method specification to markdown."
        )
        autogen.agentchat.register_function(
            partial(write_proposed_method_spec_json, artifact_dir=artifact_dir / "pc_agents" / "researcher"),
            caller=agent, executor=agent, name="write_proposed_method_spec_json",
            description="Persist the proposed method specification to JSON."
        )

    def _register_pc_generic_tools(agent: autogen.ConversableAgent, slug: str, mu_vector: np.ndarray) -> None:
        """Binds tools for the PC Generic Agents."""
        autogen.agentchat.register_function(
            partial(run_optimizer, method=slug, sigma=sigma, mu=mu_vector, rf=rf, benchmark_weights=benchmark_weights, constraints=ips_constraints, returns_matrix=returns_matrix),
            caller=agent, executor=agent, name="run_optimizer",
            description="Execute the portfolio optimization algorithm."
        )
        autogen.agentchat.register_function(
            partial(check_ips_compliance, sigma=sigma, benchmark_weights=benchmark_weights, constraints=ips_constraints),
            caller=agent, executor=agent, name="check_ips_compliance",
            description="Verify that the proposed weight vector satisfies all IPS constraints."
        )
        autogen.agentchat.register_function(
            partial(write_pc_weights_json, method=slug, artifact_dir=artifact_dir / "pc_agents"),
            caller=agent, executor=agent, name="write_pc_weights_json",
            description="Serialize and persist the optimized portfolio weights to JSON."
        )
        autogen.agentchat.register_function(
            partial(write_pc_report_md, method=slug, artifact_dir=artifact_dir / "pc_agents", benchmark_weights=benchmark_weights.tolist(), macro_view=macro_view_data),
            caller=agent, executor=agent, name="write_pc_report_md",
            description="Format and persist the human-readable portfolio construction report."
        )

    for method_slug, method_config in pc_registry.items():
        if method_slug == "adversarial_diversifier":
            continue

        template_key: str = method_config["prompt_template_key"]
        raw_prompt: str = prompt_templates[template_key]["content"]

        if method_slug == "researcher":
            sys_msg: str = raw_prompt.format(
                PC_METHOD_REGISTRY_PATH=str(artifact_dir / "method_registry.json"),
                IPS_TEXT_OR_PATH="IPS.md",
                MACRO_VIEW_PATH=str(macro_view_path),
                TOOL_LITERATURE_SEARCH="tool_literature_search",
                PY_VALIDATE_METHOD_SPEC="validate_method_spec",
                PROPOSED_METHOD_OUTPUT_PATH="proposed_method.md",
                PROPOSED_METHOD_SPEC_OUTPUT_PATH="proposed_method_spec.json",
            )
            _term_fn = _make_termination_fn(react_params["ac_equity_agent"]["termination_tools"])

            agent = autogen.ConversableAgent(
                name="pc_researcher", system_message=sys_msg,
                llm_config=_create_llm_config(method_config["llm_policy_key"], config, openai_api_key),
                human_input_mode="NEVER", is_termination_msg=_term_fn,
            )
            _register_researcher_tools(agent)
            agents["Phase_A"]["researcher"] = agent

            def _researcher_factory(_sys=sys_msg, _llm=_create_llm_config(method_config["llm_policy_key"], config, openai_api_key), _tf=_term_fn) -> autogen.ConversableAgent:
                _a = autogen.ConversableAgent(name="pc_researcher", system_message=_sys, llm_config=copy.deepcopy(_llm), human_input_mode="NEVER", is_termination_msg=_tf)
                _register_researcher_tools(_a)
                return _a
            agent_factories["researcher"] = _researcher_factory

        else:
            uses_cmas: bool = method_config.get("uses_cmas", False)
            if uses_cmas and method_slug not in mu_dict:
                raise ValueError(f"mu_dict is missing a required entry for method '{method_slug}' (uses_cmas=True).")

            mu_vector: np.ndarray = mu_dict[method_slug] if uses_cmas else mu_dict.get(method_slug, np.zeros(18, dtype=np.float64))

            sys_msg = raw_prompt.format(
                PC_METHOD_NAME=method_slug, PC_FAMILY=method_config["family"], USES_CMAS=str(uses_cmas),
                IPS_TEXT_OR_PATH="IPS.md", CMA_DIR=str(cma_dir), COV_MATRIX_PATH="sigma.npy",
                MACRO_VIEW_PATH=str(macro_view_path), BENCHMARK_DEF_PATH="benchmark.json",
                PY_RUN_PC_METHOD="run_optimizer", PY_CHECK_IPS_COMPLIANCE="check_ips_compliance",
                PY_WRITE_PC_WEIGHTS_JSON="write_pc_weights_json", PY_WRITE_PC_REPORT_MD="write_pc_report_md",
                PC_WEIGHTS_OUTPUT_PATH="pc_weights.json", PC_REPORT_OUTPUT_PATH="pc_report.md",
            )
            _term_fn = _make_termination_fn(["write_pc_weights_json", "write_pc_report_md"])

            agent = autogen.ConversableAgent(
                name=f"pc_{method_slug}", system_message=sys_msg,
                llm_config=_create_llm_config(method_config["llm_policy_key"], config, openai_api_key),
                human_input_mode="NEVER", is_termination_msg=_term_fn,
            )
            _register_pc_generic_tools(agent, method_slug, mu_vector)
            agents["Phase_A"][method_slug] = agent

            def _pc_factory(_slug=method_slug, _sys=sys_msg, _llm=_create_llm_config(method_config["llm_policy_key"], config, openai_api_key), _tf=_term_fn, _mu=mu_vector) -> autogen.ConversableAgent:
                _a = autogen.ConversableAgent(name=f"pc_{_slug}", system_message=_sys, llm_config=copy.deepcopy(_llm), human_input_mode="NEVER", is_termination_msg=_tf)
                _register_pc_generic_tools(_a, _slug, _mu)
                return _a
            agent_factories[method_slug] = _pc_factory

    # ==================================================================
    # PHASE B: PC ADVERSARIAL DIVERSIFIER AGENT
    # ==================================================================
    logger.info("Instantiating Phase B: Adversarial Diversifier agent.")

    def _register_adversarial_tools(agent: autogen.ConversableAgent, mu_ms: np.ndarray) -> None:
        """Binds tools for the Adversarial Diversifier Agent."""
        autogen.agentchat.register_function(
            partial(compute_ensemble_centroid, other_pc_weights_dir=artifact_dir / "pc_agents"),
            caller=agent, executor=agent, name="compute_ensemble_centroid",
            description="Compute the equal-weighted centroid of all non-adversarial PC portfolios."
        )
        autogen.agentchat.register_function(
            partial(run_adversarial_diversifier_optimizer, sigma=sigma, mu=mu_ms, rf=rf, benchmark_weights=benchmark_weights, constraints=ips_constraints),
            caller=agent, executor=agent, name="run_adversarial_diversifier_optimizer",
            description="Solve the adversarial diversifier optimization problem."
        )
        autogen.agentchat.register_function(
            partial(check_sharpe_floor, max_sharpe_diagnostics_path=artifact_dir / "pc_agents" / "max_sharpe" / "pc_weights.json", returns_matrix=returns_matrix, rf=rf),
            caller=agent, executor=agent, name="check_sharpe_floor",
            description="Compute the Sharpe floor for the adversarial constraint."
        )
        autogen.agentchat.register_function(
            partial(check_ips_compliance, sigma=sigma, benchmark_weights=benchmark_weights, constraints=ips_constraints),
            caller=agent, executor=agent, name="check_ips_compliance",
            description="Verify that the proposed weight vector satisfies all IPS constraints."
        )
        autogen.agentchat.register_function(
            partial(write_pc_weights_json, method="adversarial_diversifier", artifact_dir=artifact_dir / "pc_agents"),
            caller=agent, executor=agent, name="write_pc_weights_json",
            description="Serialize and persist the optimized portfolio weights to JSON."
        )
        autogen.agentchat.register_function(
            partial(write_pc_report_md, method="adversarial_diversifier", artifact_dir=artifact_dir / "pc_agents", benchmark_weights=benchmark_weights.tolist(), macro_view=macro_view_data),
            caller=agent, executor=agent, name="write_pc_report_md",
            description="Format and persist the human-readable portfolio construction report."
        )

    adv_config: Dict[str, Any] = pc_registry["adversarial_diversifier"]
    adv_sys_msg: str = prompt_templates[adv_config["prompt_template_key"]]["content"].format(
        IPS_TEXT_OR_PATH="IPS.md", CMA_DIR=str(cma_dir), COV_MATRIX_PATH="sigma.npy",
        OTHER_PC_WEIGHTS_DIR=str(artifact_dir / "pc_agents"),
        MAX_SHARPE_DIAGNOSTICS_PATH=str(artifact_dir / "pc_agents" / "max_sharpe" / "pc_weights.json"),
        PY_COMPUTE_ENSEMBLE_CENTROID="compute_ensemble_centroid",
        PY_RUN_ADVERSARIAL_DIVERSIFIER_OPTIMIZER="run_adversarial_diversifier_optimizer",
        PY_CHECK_SHARPE_FLOOR="check_sharpe_floor", PY_CHECK_IPS_COMPLIANCE="check_ips_compliance",
        PY_WRITE_PC_WEIGHTS_JSON="write_pc_weights_json", PY_WRITE_PC_REPORT_MD="write_pc_report_md",
        PC_WEIGHTS_OUTPUT_PATH="pc_weights.json", PC_REPORT_OUTPUT_PATH="pc_report.md",
    )
    _adv_term_fn = _make_termination_fn(["write_pc_weights_json", "write_pc_report_md"])

    adv_agent = autogen.ConversableAgent(
        name="pc_adversarial_agent", system_message=adv_sys_msg,
        llm_config=_create_llm_config(adv_config["llm_policy_key"], config, openai_api_key),
        human_input_mode="NEVER", is_termination_msg=_adv_term_fn,
    )
    max_sharpe_mu: np.ndarray = mu_dict.get("max_sharpe", np.zeros(18, dtype=np.float64))
    _register_adversarial_tools(adv_agent, max_sharpe_mu)
    agents["Phase_B"]["adversarial_diversifier"] = adv_agent

    def _adv_factory(_sys=adv_sys_msg, _llm=_create_llm_config(adv_config["llm_policy_key"], config, openai_api_key), _tf=_adv_term_fn, _mu_ms=max_sharpe_mu) -> autogen.ConversableAgent:
        _a = autogen.ConversableAgent(name="pc_adversarial_agent", system_message=_sys, llm_config=copy.deepcopy(_llm), human_input_mode="NEVER", is_termination_msg=_tf)
        _register_adversarial_tools(_a, _mu_ms)
        return _a
    agent_factories["adversarial_diversifier"] = _adv_factory

    # ==================================================================
    # PHASE C: CRO AGENT TEMPLATE
    # ==================================================================
    logger.info("Instantiating Phase C: CRO Agent template.")
    cro_config: Dict[str, Any] = config["REGISTRIES"]["CRO_AGENT"]
    cro_prompt_template: str = prompt_templates[cro_config["prompt_template_key"]]["content"]
    _cro_term_fn = _make_termination_fn(["write_cro_report_json", "write_cro_narrative_md"])

    cro_agent_template = autogen.ConversableAgent(
        name="cro_agent_template", system_message="PLACEHOLDER",
        llm_config=_create_llm_config(cro_config["llm_policy_key"], config, openai_api_key),
        human_input_mode="NEVER", is_termination_msg=_cro_term_fn,
    )
    agents["Phase_C"]["cro_agent"] = cro_agent_template

    def _cro_factory(candidate_slug: str, _template=cro_prompt_template, _llm=_create_llm_config(cro_config["llm_policy_key"], config, openai_api_key), _tf=_cro_term_fn) -> autogen.ConversableAgent:
        formatted_sys: str = _template.format(
            CANDIDATE_ID=candidate_slug,
            PC_WEIGHTS_PATH=str(artifact_dir / "pc_agents" / candidate_slug / "pc_weights.json"),
            IPS_TEXT_OR_PATH="IPS.md", BENCHMARK_DEF_PATH="benchmark.json", SIGMA_PATH="sigma.npy",
            RETURNS_MATRIX_PATH="returns_matrix.parquet", FACTORS_PATH="factors.csv",
            BENCHMARK_WEIGHTS_PATH="benchmark_weights.npy", RF=str(rf)
        )
        _a = autogen.ConversableAgent(name=f"cro_{candidate_slug}", system_message=formatted_sys, llm_config=copy.deepcopy(_llm), human_input_mode="NEVER", is_termination_msg=_tf)

        # CRITICAL REMEDIATION: CRO Data Starvation Defect
        # Dynamically load the weights from disk at dispatch time to bind to the risk tools.
        # This prevents the LLM from hallucinating the 18-element array.
        weights_path: Path = artifact_dir / "pc_agents" / candidate_slug / "pc_weights.json"
        if not weights_path.exists():
            raise FileNotFoundError(f"Fatal Error: Weights file missing for {candidate_slug} at {weights_path}")

        with open(weights_path, "r", encoding="utf-8") as f:
            w_data: Dict[str, Any] = json.load(f)
        weights: List[float] = w_data["weights"]

        # Register CRO tools with weights securely bound
        autogen.agentchat.register_function(partial(compute_ex_ante_vol, weights=weights, sigma=sigma), caller=_a, executor=_a, name="compute_ex_ante_vol", description="Compute annualized ex-ante portfolio volatility.")
        autogen.agentchat.register_function(partial(compute_tracking_error, weights=weights, benchmark_weights=benchmark_weights, sigma=sigma), caller=_a, executor=_a, name="compute_tracking_error", description="Compute ex-ante tracking error vs 60/40 benchmark.")
        autogen.agentchat.register_function(partial(compute_backtest_sharpe, weights=weights, returns_matrix=returns_matrix, rf=rf), caller=_a, executor=_a, name="compute_backtest_sharpe", description="Compute annualized backtest Sharpe ratio.")
        autogen.agentchat.register_function(partial(compute_mdd, weights=weights, returns_matrix=returns_matrix), caller=_a, executor=_a, name="compute_mdd", description="Compute maximum drawdown over the backtest period.")
        autogen.agentchat.register_function(
            partial(run_factor_regression, weights=weights, returns_matrix=returns_matrix, returns_index=returns_index, factors_df=factors_df, rf=rf),
            caller=_a, executor=_a, name="run_factor_regression",
            description="Run Fama-French 3-factor OLS regression.",
        )
        autogen.agentchat.register_function(partial(check_ips_compliance, sigma=sigma, benchmark_weights=benchmark_weights, constraints=ips_constraints), caller=_a, executor=_a, name="check_ips_compliance", description="Verify weights satisfy all IPS constraints.")
        autogen.agentchat.register_function(partial(write_cro_report_json, candidate_id=candidate_slug, artifact_dir=artifact_dir / "cro_reports"), caller=_a, executor=_a, name="write_cro_report_json", description="Persist CRO risk metrics as JSON artifact.")
        autogen.agentchat.register_function(partial(write_cro_narrative_md, candidate_id=candidate_slug, artifact_dir=artifact_dir / "cro_reports"), caller=_a, executor=_a, name="write_cro_narrative_md", description="Persist CRO risk narrative as markdown.")

        return _a
    agent_factories["cro"] = _cro_factory

    # ==================================================================
    # PHASE D: PEER REVIEW AGENT TEMPLATE
    # ==================================================================
    logger.info("Instantiating Phase D: Peer Review Agent template.")

    def _register_peer_tools(agent: autogen.ConversableAgent) -> None:
        """Binds tools for the Peer Review Agent."""
        autogen.agentchat.register_function(
            partial(write_peer_review_md, artifact_dir=artifact_dir / "peer_reviews_staging"),
            caller=agent, executor=agent, name="write_peer_review_md",
            description="Format and persist a peer review to a markdown artifact."
        )

    peer_config: Dict[str, Any] = config["REGISTRIES"]["PEER_REVIEW"]
    peer_prompt_template: str = prompt_templates[peer_config["prompt_template_key"]]["content"]
    _peer_term_fn = _make_termination_fn(["write_peer_review_md"])

    peer_agent_template = autogen.ConversableAgent(
        name="peer_review_agent_template", system_message="PLACEHOLDER",
        llm_config=_create_llm_config(peer_config["llm_policy_key"], config, openai_api_key),
        human_input_mode="NEVER", is_termination_msg=_peer_term_fn,
    )
    _register_peer_tools(peer_agent_template)
    agents["Phase_D"]["peer_review_agent"] = peer_agent_template

    def _peer_factory(reviewer_slug: str, peer_slugs: List[str], _template=peer_prompt_template, _llm=_create_llm_config(peer_config["llm_policy_key"], config, openai_api_key), _tf=_peer_term_fn) -> autogen.ConversableAgent:
        formatted_sys: str = _template.format(
            REVIEWER_SLUG=reviewer_slug, PEER_SLUGS=peer_slugs,
            PEER_PC_WEIGHTS_PATH="pc_weights.json", PEER_PC_REPORT_PATH="pc_report.md",
            PEER_CRO_REPORT_PATH="cro_report.json", MACRO_VIEW_PATH="macro-view.json",
            IPS_TEXT_OR_PATH="IPS.md", OWN_PC_WEIGHTS_PATH="pc_weights.json"
        )
        _a = autogen.ConversableAgent(name=f"peer_{reviewer_slug}", system_message=formatted_sys, llm_config=copy.deepcopy(_llm), human_input_mode="NEVER", is_termination_msg=_tf)
        _register_peer_tools(_a)
        return _a
    agent_factories["peer_review"] = _peer_factory

    # ==================================================================
    # PHASE E: VOTING AGENT TEMPLATE
    # ==================================================================
    logger.info("Instantiating Phase E: Voting Agent template.")
    vote_config: Dict[str, Any] = config["REGISTRIES"]["VOTING"]
    vote_prompt_template: str = prompt_templates[vote_config["prompt_template_key"]]["content"]
    _vote_term_fn = _make_termination_fn(["submit_vote"])

    canonical_slug_list: List[str] = [s for s in pc_registry.keys() if s not in {"researcher", "adversarial_diversifier"}]
    canonical_slug_list.append("adversarial_diversifier")
    canonical_slug_set: Set[str] = set(canonical_slug_list)

    vote_agent_template = autogen.ConversableAgent(
        name="voting_agent_template", system_message="PLACEHOLDER",
        llm_config=_create_llm_config(vote_config["llm_policy_key"], config, openai_api_key, force_tool_choice=vote_config.get("forced_tool_call")),
        human_input_mode="NEVER", is_termination_msg=_vote_term_fn,
    )
    agents["Phase_E"]["voting_agent"] = vote_agent_template

    def _voting_factory(voter_slug: str, _template=vote_prompt_template, _llm=_create_llm_config(vote_config["llm_policy_key"], config, openai_api_key, force_tool_choice=vote_config.get("forced_tool_call")), _tf=_vote_term_fn, _valid_slugs=canonical_slug_set) -> autogen.ConversableAgent:
        formatted_sys: str = _template.format(
            VOTER_SLUG=voter_slug, CANDIDATE_LIST=list(_valid_slugs),
            CRO_REPORTS_PATH="cro_reports", PEER_REVIEWS_DIR="peer_reviews",
            IPS_TEXT_OR_PATH="IPS.md", MACRO_VIEW_PATH="macro-view.json", OWN_METHOD_SLUG=voter_slug
        )
        _a = autogen.ConversableAgent(name=f"voter_{voter_slug}", system_message=formatted_sys, llm_config=copy.deepcopy(_llm), human_input_mode="NEVER", is_termination_msg=_tf)

        bound_submit_vote = partial(submit_vote, voter_id=voter_slug, own_method_slug=voter_slug, valid_candidate_slugs=_valid_slugs, artifact_dir=artifact_dir / "voting")
        autogen.agentchat.register_function(bound_submit_vote, caller=_a, executor=_a, name="submit_vote", description="Submit a structured Borda-count vote.")
        return _a
    agent_factories["voting"] = _voting_factory

    # ==================================================================
    # PHASE F: TOP-5 REVISION AGENT TEMPLATE
    # ==================================================================
    logger.info("Instantiating Phase F: Top-5 Revision Agent template.")

    def _register_revision_tools(agent: autogen.ConversableAgent, slug: str, mu_vector: np.ndarray) -> None:
        """Binds tools for the Top-5 Revision Agent."""
        autogen.agentchat.register_function(
            partial(run_optimizer, method=slug, sigma=sigma, mu=mu_vector, rf=rf, benchmark_weights=benchmark_weights, constraints=ips_constraints, returns_matrix=returns_matrix),
            caller=agent, executor=agent, name="run_optimizer",
            description="Re-run the portfolio optimizer with optional constraint overrides."
        )
        autogen.agentchat.register_function(
            partial(check_ips_compliance, sigma=sigma, benchmark_weights=benchmark_weights, constraints=ips_constraints),
            caller=agent, executor=agent, name="check_ips_compliance",
            description="Verify that the revised weight vector satisfies all IPS constraints."
        )
        autogen.agentchat.register_function(
            partial(write_pc_weights_json, method=slug, artifact_dir=artifact_dir / "top5_revision"),
            caller=agent, executor=agent, name="write_pc_weights_json",
            description="Serialize and persist the revised portfolio weights to JSON."
        )
        autogen.agentchat.register_function(
            partial(write_pc_report_md, method=slug, artifact_dir=artifact_dir / "top5_revision", benchmark_weights=benchmark_weights.tolist(), macro_view=macro_view_data),
            caller=agent, executor=agent, name="write_pc_report_md",
            description="Format and persist the revised portfolio construction report."
        )
        autogen.agentchat.register_function(
            partial(confirm_no_revision, method=slug, artifact_dir=artifact_dir / "top5_revision" / slug),
            caller=agent, executor=agent, name="confirm_no_revision",
            description="Record and persist a no-revision decision."
        )

    rev_config: Dict[str, Any] = config["REGISTRIES"]["TOP5_REVISION"]
    rev_prompt_template: str = prompt_templates[rev_config["prompt_template_key"]]["content"]
    _rev_term_fn = _make_termination_fn(["write_pc_weights_json", "confirm_no_revision"])

    rev_agent_template = autogen.ConversableAgent(
        name="top5_revision_agent_template", system_message="PLACEHOLDER",
        llm_config=_create_llm_config(rev_config["llm_policy_key"], config, openai_api_key),
        human_input_mode="NEVER", is_termination_msg=_rev_term_fn,
    )
    agents["Phase_F"]["top5_revision_agent"] = rev_agent_template

    def _revision_factory(slug: str, _template=rev_prompt_template, _llm=_create_llm_config(rev_config["llm_policy_key"], config, openai_api_key), _tf=_rev_term_fn) -> autogen.ConversableAgent:
        formatted_sys: str = _template.format(
            CURRENT_PC_WEIGHTS_PATH="pc_weights.json", CURRENT_PC_REPORT_PATH="pc_report.md",
            OWN_CRO_REPORT_PATH="cro_report.json", RECEIVED_REVIEWS_DIR="peer_reviews",
            FULL_REVIEWS_DIR="peer_reviews", COMPOSITE_RANKING_PATH="composite_ranking.json",
            TOP5_SELECTION_PATH="top5_selection.json", SIGMA_PATH="sigma.npy",
            CMA_DIR="cma_judge", IPS_TEXT_OR_PATH="IPS.md",
            PY_RUN_PC_METHOD="run_optimizer", PY_CHECK_IPS_COMPLIANCE="check_ips_compliance",
            PY_WRITE_PC_WEIGHTS_JSON="write_pc_weights_json", PY_WRITE_PC_REPORT_MD="write_pc_report_md"
        )
        _a = autogen.ConversableAgent(name=f"rev_{slug}", system_message=formatted_sys, llm_config=copy.deepcopy(_llm), human_input_mode="NEVER", is_termination_msg=_tf)

        mu_vector: np.ndarray = mu_dict.get(slug, np.zeros(18, dtype=np.float64))
        _register_revision_tools(_a, slug, mu_vector)
        return _a
    agent_factories["revision"] = _revision_factory

    # ==================================================================
    # PHASE G: CIO AGENT
    # ==================================================================
    logger.info("Instantiating Phase G: CIO Agent.")

    def _register_cio_tools(agent: autogen.ConversableAgent, state: Dict[str, Any]) -> None:
        """
        Binds tools for the CIO Agent.
        CRITICAL REMEDIATION: Uses wrapper functions to dynamically pull `top5_slugs`
        from the `pipeline_state` dictionary at runtime, resolving the temporal paradox.
        Also implements the `write_board_memo_wrapper` to prevent JSON escaping crashes.
        """
        def wrapper_compute_cio_composite_scores() -> Dict[str, Any]:
            top5: List[str] = state.get("top5_slugs", [])
            if not top5:
                raise ValueError("Fatal Error: top5_slugs not found in pipeline_state. Phase F must complete first.")
            return compute_cio_composite_scores(
                candidate_methods=top5, artifact_dir=artifact_dir, sigma=sigma,
                benchmark_weights=benchmark_weights, returns_matrix=returns_matrix,
                rf=rf, macro_view=macro_view_data, cma_set=cma_set,
                universe_map=config["REGISTRIES"]["universe_map"], canonical_order=canonical_order
            )

        def wrapper_evaluate_ensemble_methods(composite_scores: Dict[str, float]) -> Dict[str, List[float]]:
            top5: List[str] = state.get("top5_slugs", [])
            if not top5:
                raise ValueError("Fatal Error: top5_slugs not found in pipeline_state.")
            return evaluate_ensemble_methods(
                top_5_methods=top5, artifact_dir=artifact_dir, sigma=sigma,
                benchmark_weights=benchmark_weights, returns_matrix=returns_matrix,
                rf=rf, composite_scores=composite_scores, macro_view=macro_view_data,
                max_weight_per_asset=ips_constraints.get("max_weight_per_asset", 0.25)
            )

        def wrapper_check_ips_compliance_ensemble(ensemble_weights: Dict[str, List[float]]) -> Dict[str, Dict[str, Any]]:
            return check_ips_compliance_ensemble(
                ensemble_weights=ensemble_weights, sigma=sigma,
                benchmark_weights=benchmark_weights, constraints=ips_constraints
            )

        def wrapper_select_best_ensemble(ensemble_scores: Dict[str, float], compliance_flags: Dict[str, Dict[str, Any]], ensemble_weights: Dict[str, List[float]]) -> Dict[str, Any]:
            result: Dict[str, Any] = select_best_ensemble(ensemble_scores, compliance_flags, ensemble_weights)
            # Save the selected weights and ensemble name to the pipeline state for the board memo wrapper
            state["final_weights"] = result["weights"]
            state["selected_ensemble"] = result["selected_ensemble"]
            return result

        def write_board_memo_wrapper(rationale: str, board_memo_content: str) -> str:
            final_weights: List[float] = state.get("final_weights", [])
            selected_ensemble: str = state.get("selected_ensemble", "")
            if not final_weights or not selected_ensemble:
                raise ValueError("Fatal Error: final_weights or selected_ensemble missing from pipeline_state.")

            # Persist the JSON artifact
            write_final_weights_json(
                weights=final_weights, rationale=rationale, selected_ensemble=selected_ensemble,
                artifact_dir=artifact_dir / "cio", asset_class_labels=canonical_order
            )
            # Persist the Markdown artifact
            return write_board_memo_md(
                weights=final_weights, rationale=board_memo_content, macro_summary=macro_view_data,
                cma_summary=cma_set, risk_summary={}, selected_ensemble=selected_ensemble,
                artifact_dir=artifact_dir / "cio", asset_class_labels=canonical_order,
                benchmark_weights=benchmark_weights.tolist()
            )

        # Register the wrappers
        autogen.agentchat.register_function(wrapper_compute_cio_composite_scores, caller=agent, executor=agent, name="compute_cio_composite_scores", description="Compute the CIO multi-criteria composite scores for the top-5 candidates.")
        autogen.agentchat.register_function(wrapper_evaluate_ensemble_methods, caller=agent, executor=agent, name="evaluate_ensemble_methods", description="Evaluate all seven ensemble combination methods.")
        autogen.agentchat.register_function(wrapper_check_ips_compliance_ensemble, caller=agent, executor=agent, name="check_ips_compliance_ensemble", description="Check IPS compliance for each of the seven ensemble weight vectors.")
        autogen.agentchat.register_function(wrapper_select_best_ensemble, caller=agent, executor=agent, name="select_best_ensemble", description="Select the best IPS-compliant ensemble based on the CIO composite scores.")
        autogen.agentchat.register_function(write_board_memo_wrapper, caller=agent, executor=agent, name="write_board_memo", description="Persist the executive rationale and the full Board Memo markdown.")

    cio_config: Dict[str, Any] = config["REGISTRIES"]["CIO_AGENT"]
    cio_prompt_template: str = prompt_templates[cio_config["prompt_template_key"]]["content"]
    # Note: The termination tool is now the wrapper 'write_board_memo'
    _cio_term_fn = _make_termination_fn(["write_board_memo"])

    cio_sys_msg: str = cio_prompt_template.format(
        PC_CANDIDATES_DIR="top5_revision", CRO_REPORTS_PATH="cro_reports",
        PEER_REVIEWS_DIR="peer_reviews", VOTE_TALLIES_PATH="vote_corpus.json",
        METRIC_SCORES_PATH="composite_ranking.json", IPS_TEXT_OR_PATH="IPS.md",
        MACRO_VIEW_PATH="macro-view.json", BENCHMARK_DEF_PATH="benchmark.json",
        SIGMA_PATH="sigma.npy", RETURNS_MATRIX_PATH="returns_matrix.parquet",
        BENCHMARK_WEIGHTS_PATH="benchmark_weights.npy", RF=str(rf), CMA_JUDGE_DIR="cma_judge",
        PY_COMPUTE_CIO_COMPOSITE_SCORES="compute_cio_composite_scores",
        PY_EVALUATE_ENSEMBLE_METHODS="evaluate_ensemble_methods",
        PY_CHECK_IPS_COMPLIANCE_ENSEMBLE="check_ips_compliance_ensemble",
        PY_SELECT_BEST_ENSEMBLE="select_best_ensemble",
        PY_WRITE_FINAL_WEIGHTS_JSON="write_board_memo", # Replaced with wrapper
        PY_WRITE_BOARD_MEMO_MD="write_board_memo"       # Replaced with wrapper
    )

    cio_agent = autogen.ConversableAgent(
        name="cio_agent", system_message=cio_sys_msg,
        llm_config=_create_llm_config(cio_config["llm_policy_key"], config, openai_api_key),
        human_input_mode="NEVER", is_termination_msg=_cio_term_fn,
    )
    _register_cio_tools(cio_agent, pipeline_state)
    agents["Phase_G"]["cio_agent"] = cio_agent

    # ==================================================================
    # FSM CONTROLLER AND GROUPCHAT
    # ==================================================================
    all_groupchat_participants: List[autogen.ConversableAgent] = []
    for phase_agents in agents.values():
        all_groupchat_participants.extend(phase_agents.values())

    all_agents_by_name: Dict[str, autogen.ConversableAgent] = {a.name: a for a in all_groupchat_participants}

    _system_orchestrator_name: str = "System_Orchestrator"
    phase_speaker_map: Dict[str, List[str]] = {
        "A": [_system_orchestrator_name], "B": ["pc_adversarial_agent"],
        "C": [_system_orchestrator_name], "D": [_system_orchestrator_name],
        "E": [_system_orchestrator_name], "F": [_system_orchestrator_name],
        "G": ["cio_agent"],
    }

    fsm_controller = _StrategyReviewFSMController(
        phase_speaker_map=phase_speaker_map, artifact_dir=artifact_dir,
        validate_artifact_fn=validate_artifact_fn, all_agents_by_name=all_agents_by_name, config=config,
    )

    group_chat = autogen.GroupChat(
        agents=all_groupchat_participants, messages=[],
        max_round=config["INFRASTRUCTURE"]["AUTOGEN_TOPOLOGY"]["groupchat_max_round"],
        speaker_selection_method=fsm_controller,
    )

    group_chat_manager = autogen.GroupChatManager(groupchat=group_chat, llm_config=False)
    all_agents_by_name[_system_orchestrator_name] = group_chat_manager

    logger.info("MAS topology instantiation complete.")

    return MASTopology(
        agents=agents, group_chat=group_chat, group_chat_manager=group_chat_manager,
        fsm_controller=fsm_controller, agent_factories=agent_factories,
        cro_prompt_template=cro_prompt_template, pc_registry=pc_registry, mu_dict=mu_dict,
        pipeline_state=pipeline_state
    )

# -------------------------------------------------------------------------------------------------------------------------------
# Task 19, Step 2 — Execute Phase A (PC Weight Submission)
# -------------------------------------------------------------------------------------------------------------------------------

def execute_phase_a_pc_submission(
    phase_a_agents:       Dict[str, autogen.ConversableAgent],
    user_proxy:           autogen.UserProxyAgent,
    artifact_dir:         Path,
    validate_artifact_fn: Callable[[Path, str], bool],
    config:               Dict[str, Any],
    openai_api_key:       str,
    agent_factories:      Dict[str, Any],
    max_workers:          int = 8,
) -> None:
    """
    Executes Phase A: dispatches all 21 non-adversarial PC agents concurrently.
    Each thread receives a freshly instantiated ConversableAgent and a freshly
    instantiated UserProxyAgent to eliminate shared-state race conditions.
    Enforces fail-closed schema validation on all output artifacts.

    Corrections versus prior draft
    --------------------------------
    * Fresh ConversableAgent per thread via agent_factories.
    * Fresh UserProxyAgent per thread.
    * max_turns sourced from REACT_LOOP_PARAMS (not hard-coded to 1).
    * SchemaValidationError raised (defined in this module).
    * Researcher's output artifacts validated against their schema.
    * Count assertion: exactly 21 artifacts expected.

    Parameters
    ----------
    phase_a_agents : Dict[str, autogen.ConversableAgent]
        Template agents — NOT used for execution; provide slug enumeration only.
    user_proxy : autogen.UserProxyAgent
        Template proxy — NOT shared across threads.  Referenced for config only.
    artifact_dir : Path
        Root artifact directory.
    validate_artifact_fn : Callable[[Path, str], bool]
        Bound schema-validation callable.
    config : Dict[str, Any]
        STUDY_CONFIG dictionary for REACT_LOOP_PARAMS extraction.
    openai_api_key : str
        API key for fresh agent instantiation within threads.
    agent_factories : Dict[str, Any]
        Factory callables keyed by method slug.
    max_workers : int, optional
        Maximum concurrent threads.  Defaults to 8.

    Raises
    ------
    RuntimeError
        If any agent fails execution.
    SchemaValidationError
        If any produced artifact fails schema validation (fail-closed).
    """
    logger.info("Phase A: Dispatching %d PC agents concurrently.", len(phase_a_agents))

    # Extract per-agent max_turns from the frozen REACT_LOOP_PARAMS configuration
    react_params: Dict[str, Any] = config["INFRASTRUCTURE"]["REACT_LOOP_PARAMS"]
    # Researcher uses a different iteration budget than generic PC agents
    researcher_max_turns:  int = react_params.get("ac_equity_agent", {}).get("max_iterations", 30)
    generic_max_turns:     int = react_params.get("ac_equity_agent", {}).get("max_iterations", 30)

    def _run_pc_agent(slug: str) -> None:
        """Inner function executed by each worker thread for one PC agent."""
        # Instantiate a fresh ConversableAgent from the factory — no shared state
        fresh_agent: autogen.ConversableAgent = agent_factories[slug]()
        # Instantiate a fresh UserProxyAgent to prevent cross-thread state contamination
        fresh_proxy = autogen.UserProxyAgent(
            name=f"proxy_{slug}",
            human_input_mode="NEVER",
            code_execution_config=False,
        )
        # Determine the appropriate iteration budget for this agent type
        _max_turns: int = (
            researcher_max_turns if slug == "researcher" else generic_max_turns
        )
        # Compose the dispatch message with the method slug for agent orientation
        dispatch_msg: str = (
            f"Begin portfolio construction for method '{slug}'. "
            f"Execute the prescribed workflow strictly in sequential order. "
            f"Call all required tools before writing output artifacts."
        )
        # Initiate the ReAct chat between the fresh proxy and the fresh agent
        fresh_proxy.initiate_chat(
            recipient=fresh_agent,
            message=dispatch_msg,
            # max_turns governs the number of back-and-forth conversational rounds
            max_turns=_max_turns,
        )

    # Dispatch all PC agents concurrently via a thread pool
    with concurrent.futures.ThreadPoolExecutor(max_workers=max_workers) as executor:
        futures: Dict[concurrent.futures.Future, str] = {
            executor.submit(_run_pc_agent, slug): slug
            for slug in phase_a_agents.keys()
        }
        # Collect results; raise on first failure (fail-fast within phase)
        for future in concurrent.futures.as_completed(futures):
            slug_done: str = futures[future]
            try:
                future.result()
                logger.debug("Phase A: agent '%s' completed successfully.", slug_done)
            except Exception as exc:
                raise RuntimeError(
                    f"Phase A: execution failed for agent '{slug_done}': {exc}"
                ) from exc

    # ---- Schema validation gate ----
    logger.info("Phase A: Applying fail-closed schema validation gate.")
    failed_slugs: List[str] = []
    for slug in phase_a_agents.keys():
        if slug == "researcher":
            # Validate researcher's proposed method spec JSON
            spec_path: Path = artifact_dir / "pc_agents" / "researcher" / "proposed_method_spec.json"
            try:
                validate_artifact_fn(spec_path, "proposed_method_spec")
            except Exception as exc:
                failed_slugs.append(f"researcher:{exc}")
        else:
            # Validate generic PC weights JSON for every non-researcher agent
            weights_path: Path = artifact_dir / "pc_agents" / slug / "pc_weights.json"
            try:
                validate_artifact_fn(weights_path, "pc_weights")
            except Exception as exc:
                failed_slugs.append(f"{slug}:{exc}")

    if failed_slugs:
        raise SchemaValidationError(
            f"Phase A: {len(failed_slugs)} artifact(s) failed validation "
            f"(fail-closed).\nFailed: {failed_slugs}"
        )

    # Assert the exact number of pc_weights.json artifacts produced equals 20
    # (researcher produces a different artifact type; not counted here)
    produced_weights_count: int = sum(
        1 for slug in phase_a_agents.keys() if slug != "researcher"
    )
    assert produced_weights_count == 20, (
        f"Phase A count mismatch: expected 20 pc_weights.json artifacts, "
        f"found {produced_weights_count}."
    )
    logger.info(
        "Phase A: All %d artifacts passed schema validation.", len(phase_a_agents)
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 19, Step 3 — Execute Phase B (Adversarial Diversifier)
# -------------------------------------------------------------------------------------------------------------------------------

def execute_phase_b_adversarial_diversifier(
    adv_agent:            autogen.ConversableAgent,
    user_proxy:           autogen.UserProxyAgent,
    artifact_dir:         Path,
    validate_artifact_fn: Callable[[Path, str], bool],
    config:               Dict[str, Any],
    agent_factories:      Dict[str, Any],
) -> None:
    """
    Executes Phase B: dispatches the Adversarial Diversifier agent after
    verifying that exactly 20 Phase A pc_weights.json artifacts are present
    and valid, then performing fail-closed schema validation on the output.

    Corrections versus prior draft
    --------------------------------
    * Pre-condition check: verifies exactly 20 Phase A artifacts exist before
      activating the adversarial agent (enforces the sequencing constraint).
    * Fresh ConversableAgent and UserProxyAgent instantiated (no shared state).
    * max_turns sourced from REACT_LOOP_PARAMS.
    * Dispatch message now includes the concrete centroid availability hint.

    Parameters
    ----------
    adv_agent : autogen.ConversableAgent
        Template adversarial agent.
    user_proxy : autogen.UserProxyAgent
        Template proxy — NOT used for execution.
    artifact_dir : Path
        Root artifact directory.
    validate_artifact_fn : Callable[[Path, str], bool]
        Bound schema-validation callable.
    config : Dict[str, Any]
        STUDY_CONFIG dictionary.
    agent_factories : Dict[str, Any]
        Factory callables; "adversarial_diversifier" key required.

    Raises
    ------
    RuntimeError
        If Phase A pre-condition check fails or if agent execution fails.
    SchemaValidationError
        If the produced artifact fails schema validation.
    """
    logger.info("Phase B: Pre-condition check — verifying Phase A completion.")

    # Verify that exactly 20 non-adversarial pc_weights.json artifacts exist
    pc_agents_dir: Path = artifact_dir / "pc_agents"
    phase_a_artifacts: List[Path] = [
        p for p in pc_agents_dir.glob("*/pc_weights.json")
        if p.parent.name != "adversarial_diversifier"
    ]
    if len(phase_a_artifacts) != 20:
        raise RuntimeError(
            f"Phase B pre-condition failed: expected 20 Phase A pc_weights.json "
            f"artifacts, found {len(phase_a_artifacts)}.  "
            f"Phase B cannot proceed until Phase A is fully complete."
        )
    logger.info(
        "Phase B: Pre-condition passed — 20 Phase A artifacts confirmed."
    )

    # Extract Phase B max_turns from configuration
    react_params: Dict[str, Any] = config["INFRASTRUCTURE"]["REACT_LOOP_PARAMS"]
    adv_max_turns: int = react_params.get("ac_equity_agent", {}).get("max_iterations", 30)

    # Instantiate a fresh adversarial agent and proxy for this execution
    fresh_adv_agent: autogen.ConversableAgent = (
        agent_factories["adversarial_diversifier"]()
    )
    fresh_proxy = autogen.UserProxyAgent(
        name="proxy_adversarial",
        human_input_mode="NEVER",
        code_execution_config=False,
    )

    dispatch_msg: str = (
        "Begin adversarial portfolio construction.  All 20 non-adversarial "
        "pc_weights.json artifacts are confirmed present at: "
        f"{str(artifact_dir / 'pc_agents')}.  "
        "Call compute_ensemble_centroid, then check_sharpe_floor, then "
        "run_adversarial_diversifier_optimizer, then check_ips_compliance, "
        "then write_pc_weights_json, then write_pc_report_md.  "
        "Execute this sequence strictly in order."
    )

    try:
        fresh_proxy.initiate_chat(
            recipient=fresh_adv_agent,
            message=dispatch_msg,
            max_turns=adv_max_turns,
        )
    except Exception as exc:
        raise RuntimeError(
            f"Phase B: adversarial agent execution failed: {exc}"
        ) from exc

    # Schema validation gate for the adversarial artifact
    logger.info("Phase B: Applying fail-closed schema validation gate.")
    adv_weights_path: Path = (
        artifact_dir / "pc_agents" / "adversarial_diversifier" / "pc_weights.json"
    )
    try:
        validate_artifact_fn(adv_weights_path, "pc_weights")
    except Exception as exc:
        raise SchemaValidationError(
            f"Phase B: adversarial pc_weights.json failed validation: {exc}"
        ) from exc

    logger.info("Phase B: Adversarial Diversifier completed successfully.")


# -------------------------------------------------------------------------------------------------------------------------------
# Task 19, Step 4 — Execute Phase C (CRO Risk Reports)
# -------------------------------------------------------------------------------------------------------------------------------

def execute_phase_c_cro_reports(
    cro_agent:            autogen.ConversableAgent,
    user_proxy:           autogen.UserProxyAgent,
    all_pc_slugs:         List[str],
    artifact_dir:         Path,
    validate_artifact_fn: Callable[[Path, str], bool],
    config:               Dict[str, Any],
    agent_factories:      Dict[str, Any],
    max_workers:          int = 8,
) -> None:
    """
    Executes Phase C: dispatches a fresh CROAgent loop for each of the 21
    candidate portfolios concurrently.  Each thread receives its own agent
    instance with the system_message formatted for the specific candidate.
    Enforces mandatory tool-call sequence tracking and fail-closed schema
    validation.

    Corrections versus prior draft
    --------------------------------
    * Fresh CROAgent per candidate via agent_factories["cro"](candidate_slug).
    * CRO system_message formatted per candidate at dispatch time (not a broken
      unformatted template).
    * Fresh UserProxyAgent per thread.
    * max_turns sourced from REACT_LOOP_PARAMS.
    * Count assertion: exactly 21 cro_report.json artifacts expected.

    Parameters
    ----------
    cro_agent : autogen.ConversableAgent
        Template CRO agent — NOT used for execution; provides slug enumeration.
    user_proxy : autogen.UserProxyAgent
        Template proxy — NOT used for execution.
    all_pc_slugs : List[str]
        List of all 21 candidate portfolio method slugs.
    artifact_dir : Path
        Root artifact directory.
    validate_artifact_fn : Callable[[Path, str], bool]
        Bound schema-validation callable.
    config : Dict[str, Any]
        STUDY_CONFIG dictionary.
    agent_factories : Dict[str, Any]
        Factory callables; "cro" key required (accepts candidate_slug argument).
    max_workers : int, optional
        Maximum concurrent threads.  Defaults to 8.

    Raises
    ------
    RuntimeError
        If any CRO loop fails execution.
    SchemaValidationError
        If any cro_report.json fails validation (fail-closed).
    """
    logger.info(
        "Phase C: Dispatching CRO loops for %d candidates concurrently.",
        len(all_pc_slugs),
    )

    # Enforce count precondition: must have exactly 21 candidate slugs
    assert len(all_pc_slugs) == 21, (
        f"Phase C expected 21 candidate slugs, received {len(all_pc_slugs)}."
    )

    # Extract CRO agent max_turns from REACT_LOOP_PARAMS
    react_params:    Dict[str, Any] = config["INFRASTRUCTURE"]["REACT_LOOP_PARAMS"]
    cro_max_turns:   int = react_params.get("cro_narrative_agent", {}).get(
        "max_iterations", 20
    )
    # Extract mandatory tool-call sequence for the CRO agent
    cro_tool_sequence: List[str] = react_params.get(
        "cro_narrative_agent", {}
    ).get("mandatory_tool_sequence", [])

    def _run_cro_for_candidate(slug: str) -> None:
        """Inner function executed by each worker thread for one candidate."""
        # Create a fresh, per-candidate CRO agent with formatted system_message
        fresh_cro: autogen.ConversableAgent = agent_factories["cro"](slug)
        # Create a fresh proxy for this thread to prevent state contamination
        fresh_proxy = autogen.UserProxyAgent(
            name=f"proxy_cro_{slug}",
            human_input_mode="NEVER",
            code_execution_config=False,
        )
        # Compose dispatch message including the mandatory tool-call sequence
        dispatch_msg: str = (
            f"Generate the CRO risk report for candidate portfolio '{slug}'.  "
            f"You must call tools in this exact order: {cro_tool_sequence}.  "
            f"Any out-of-order call will be rejected.  "
            f"Load pc_weights.json from the pc_agents/{slug}/ directory before proceeding."
        )
        fresh_proxy.initiate_chat(
            recipient=fresh_cro,
            message=dispatch_msg,
            max_turns=cro_max_turns,
        )

    # Dispatch all 21 CRO loops concurrently
    with concurrent.futures.ThreadPoolExecutor(max_workers=max_workers) as executor:
        futures: Dict[concurrent.futures.Future, str] = {
            executor.submit(_run_cro_for_candidate, slug): slug
            for slug in all_pc_slugs
        }
        for future in concurrent.futures.as_completed(futures):
            slug_done: str = futures[future]
            try:
                future.result()
                logger.debug("Phase C: CRO report for '%s' completed.", slug_done)
            except Exception as exc:
                raise RuntimeError(
                    f"Phase C: CRO execution failed for candidate '{slug_done}': {exc}"
                ) from exc

    # Fail-closed schema validation gate for all 21 CRO report artifacts
    logger.info("Phase C: Applying fail-closed schema validation gate.")
    failed_slugs: List[str] = []
    for slug in all_pc_slugs:
        cro_path: Path = artifact_dir / "cro_reports" / slug / "cro_report.json"
        try:
            validate_artifact_fn(cro_path, "cro_report")
        except Exception as exc:
            failed_slugs.append(f"{slug}:{exc}")

    if failed_slugs:
        raise SchemaValidationError(
            f"Phase C: {len(failed_slugs)} cro_report.json artifact(s) failed "
            f"validation (fail-closed).\nFailed: {failed_slugs}"
        )

    logger.info("Phase C: All 21 CRO reports passed schema validation.")


# -------------------------------------------------------------------------------------------------------------------------------
# Task 19, Step 5 — Execute Phase D (Peer Review)
# -------------------------------------------------------------------------------------------------------------------------------

def execute_phase_d_peer_review(
    peer_agent:      autogen.ConversableAgent,
    user_proxy:      autogen.UserProxyAgent,
    all_pc_slugs:    List[str],
    pc_registry:     Dict[str, Any],
    seed:            int,
    artifact_dir:    Path,
    config:          Dict[str, Any],
    agent_factories: Dict[str, Any],
    max_workers:     int = 8,
) -> None:
    """
    Executes Phase D: computes seeded randomised peer review assignments,
    dispatches fresh reviewer agents concurrently to a staging directory,
    validates the REVIEW_COMPLETE barrier across all 42 reviews, then
    atomically moves staging artifacts to the visible peer_reviews directory.

    Corrections versus prior draft
    --------------------------------
    * Fresh peer review agent per reviewer via agent_factories["peer_review"].
    * Fresh UserProxyAgent per thread.
    * write_peer_review_md tool registered on each reviewer agent.
    * Atomic staging barrier: reviews written to peer_reviews_staging/;
      only moved to peer_reviews/ after all 42 pass the barrier check.
    * max_turns sourced from REACT_LOOP_PARAMS.
    * Dispatch message includes concrete artifact paths for each reviewer.
    * validate_artifact_fn applied to each staging review before barrier.

    Parameters
    ----------
    peer_agent : autogen.ConversableAgent
        Template peer review agent — provides slug enumeration only.
    user_proxy : autogen.UserProxyAgent
        Template proxy — NOT used for execution.
    all_pc_slugs : List[str]
        List of all 21 candidate portfolio slugs.
    pc_registry : Dict[str, Any]
        PC method registry containing family labels.
    seed : int
        Frozen random seed for deterministic assignment reproducibility.
    artifact_dir : Path
        Root artifact directory.
    config : Dict[str, Any]
        STUDY_CONFIG dictionary.
    agent_factories : Dict[str, Any]
        Factory callables; "peer_review" key required.
    max_workers : int, optional
        Maximum concurrent threads.  Defaults to 8.

    Raises
    ------
    RuntimeError
        If agent execution fails or the 42-review barrier is not met.
    SchemaValidationError
        If any review artifact fails validation.
    """
    logger.info("Phase D: Computing review assignments and dispatching reviewers.")

    # Build the list of method metadata dicts required by compute_review_assignments
    pc_methods_list: List[Dict[str, str]] = [
        {"slug": s, "family": pc_registry[s]["family"]}
        for s in all_pc_slugs
    ]
    # Staging directory for reviews (invisible until atomic move)
    staging_dir: Path = artifact_dir / "peer_reviews_staging"
    staging_dir.mkdir(parents=True, exist_ok=True)

    # Compute the seeded randomised bipartite assignment matrix
    # Returns: Dict[reviewer_slug → [peer_slug_1, peer_slug_2]]
    assignments: Dict[str, List[str]] = compute_review_assignments(
        pc_methods=pc_methods_list,
        seed=seed,
        artifact_dir=artifact_dir / "peer_reviews",
    )

    # Assert the assignment matrix covers all 21 reviewers for exactly 42 reviews total
    total_review_count: int = sum(len(v) for v in assignments.values())
    assert len(assignments) == 21, (
        f"Phase D: expected 21 reviewers in assignment matrix, "
        f"got {len(assignments)}."
    )
    assert total_review_count == 42, (
        f"Phase D: expected 42 total reviews (21 × 2), "
        f"got {total_review_count}."
    )

    # Extract max_turns from REACT_LOOP_PARAMS for the peer review agent
    react_params:     Dict[str, Any] = config["INFRASTRUCTURE"]["REACT_LOOP_PARAMS"]
    peer_max_turns:   int = react_params.get("peer_review_agent", {}).get(
        "max_iterations", 20
    )

    def _run_reviewer(reviewer_slug: str, peer_slugs: List[str]) -> None:
        """Inner function per reviewer thread."""
        # Create a fresh peer agent with system_message scoped to this reviewer
        fresh_peer: autogen.ConversableAgent = agent_factories["peer_review"](
            reviewer_slug, peer_slugs
        )
        # Create a fresh proxy for this reviewer thread
        fresh_proxy = autogen.UserProxyAgent(
            name=f"proxy_peer_{reviewer_slug}",
            human_input_mode="NEVER",
            code_execution_config=False,
        )
        dispatch_msg: str = (
            f"You are portfolio constructor '{reviewer_slug}'.  "
            f"You must write peer reviews for the following two candidates: "
            f"{peer_slugs}.  "
            f"For each peer, read the pc_report.md from "
            f"{str(artifact_dir / 'pc_agents')} and the cro_report.json from "
            f"{str(artifact_dir / 'cro_reports')}.  "
            f"Call write_peer_review_md_<peer_slug> for each peer.  "
            f"Each review must contain the string 'REVIEW_COMPLETE' at the end."
        )
        fresh_proxy.initiate_chat(
            recipient=fresh_peer,
            message=dispatch_msg,
            max_turns=peer_max_turns,
        )

    # Dispatch all 21 reviewer agents concurrently
    with concurrent.futures.ThreadPoolExecutor(max_workers=max_workers) as executor:
        futures: Dict[concurrent.futures.Future, str] = {
            executor.submit(_run_reviewer, reviewer, peers): reviewer
            for reviewer, peers in assignments.items()
        }
        for future in concurrent.futures.as_completed(futures):
            reviewer_done: str = futures[future]
            try:
                future.result()
                logger.debug("Phase D: reviewer '%s' completed.", reviewer_done)
            except Exception as exc:
                raise RuntimeError(
                    f"Phase D: reviewer '{reviewer_done}' execution failed: {exc}"
                ) from exc

    # Validate the barrier: all 42 reviews must be present, non-empty, and contain
    # REVIEW_COMPLETE before the atomic move proceeds
    logger.info("Phase D: Validating 42-review synchronisation barrier.")
    barrier_result: Dict[str, Any] = validate_review_barrier(
        assignments, staging_dir
    )
    if not barrier_result.get("all_present", False):
        raise RuntimeError(
            f"Phase D: Synchronisation barrier failed — not all 42 reviews "
            f"are present or contain REVIEW_COMPLETE.  "
            f"Details: {barrier_result}.  Pipeline halted (fail-closed)."
        )

    # Atomic move from staging directory to the visible peer_reviews directory
    visible_dir: Path = artifact_dir / "peer_reviews"
    if visible_dir.exists():
        # Remove any stale visible directory before the atomic rename
        shutil.rmtree(visible_dir)
    # os.rename is atomic on POSIX systems (same filesystem); shutil.move is used
    # for cross-filesystem portability
    shutil.move(str(staging_dir), str(visible_dir))

    logger.info(
        "Phase D: All 42 peer reviews validated and atomically released to '%s'.",
        visible_dir,
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 19, Step 6 — Execute Phase E (Voting and Aggregation)
# -------------------------------------------------------------------------------------------------------------------------------

def execute_phase_e_voting_and_ranking(
    vote_agent:           autogen.ConversableAgent,
    user_proxy:           autogen.UserProxyAgent,
    all_pc_slugs:         List[str],
    alpha:                float,
    regime:               str,
    artifact_dir:         Path,
    validate_artifact_fn: Callable[[Path, str], bool],
    config:               Dict[str, Any],
    agent_factories:      Dict[str, Any],
    max_workers:          int = 8,
) -> Dict[str, Any]:
    """
    Executes Phase E: dispatches all 21 PC agents as voters concurrently
    with forced tool_choice=submit_vote to produce structured Borda ballots.
    Aggregates votes, extracts metric scores, and computes the composite
    ranking.

    Corrections versus prior draft
    --------------------------------
    * Fresh voting agent per voter via agent_factories["voting"] — tool_choice
      forced to submit_vote via llm_config injection in the factory.
    * Fresh UserProxyAgent per thread.
    * max_turns sourced from REACT_LOOP_PARAMS.
    * Vote count assertion: exactly 21 vote_*.json files required.
    * vote_corpus count asserted before Borda aggregation.
    * alpha sourced from config METHODOLOGY_PARAMS (not hard-coded in caller).

    Parameters
    ----------
    vote_agent : autogen.ConversableAgent
        Template voting agent — NOT used for execution.
    user_proxy : autogen.UserProxyAgent
        Template proxy — NOT used for execution.
    all_pc_slugs : List[str]
        List of all 21 candidate portfolio slugs (voters and candidates).
    alpha : float
        Regime-dependent Borda/metric blending weight from config.
    regime : str
        Current macro regime label (for composite ranking logging).
    artifact_dir : Path
        Root artifact directory.
    validate_artifact_fn : Callable[[Path, str], bool]
        Bound schema-validation callable.
    config : Dict[str, Any]
        STUDY_CONFIG dictionary.
    agent_factories : Dict[str, Any]
        Factory callables; "voting" key required.
    max_workers : int, optional
        Maximum concurrent threads.  Defaults to 8.

    Returns
    -------
    Dict[str, Any]
        Composite ranking dictionary as returned by compute_composite_ranking.

    Raises
    ------
    RuntimeError
        If any voter execution fails or the vote count is incorrect.
    SchemaValidationError
        If any vote artifact fails schema validation.
    """
    logger.info(
        "Phase E: Dispatching %d voters concurrently (regime='%s', alpha=%.2f).",
        len(all_pc_slugs), regime, alpha,
    )

    # Extract voting agent max_turns from REACT_LOOP_PARAMS
    react_params:    Dict[str, Any] = config["INFRASTRUCTURE"]["REACT_LOOP_PARAMS"]
    vote_max_turns:  int = react_params.get("voting_agent", {}).get("max_iterations", 10)
    # Ensure the voting directory exists before any voter writes to it
    (artifact_dir / "voting").mkdir(parents=True, exist_ok=True)

    def _run_voter(voter_slug: str) -> None:
        """Inner function per voter thread."""
        # Create a fresh voting agent with tool_choice=submit_vote injected
        fresh_voter: autogen.ConversableAgent = agent_factories["voting"](voter_slug)
        # Create a fresh proxy for this voter thread
        fresh_proxy = autogen.UserProxyAgent(
            name=f"proxy_voter_{voter_slug}",
            human_input_mode="NEVER",
            code_execution_config=False,
        )
        dispatch_msg: str = (
            f"You are '{voter_slug}'.  Review the peer reviews in "
            f"{str(artifact_dir / 'peer_reviews')} and the CRO reports in "
            f"{str(artifact_dir / 'cro_reports')}.  "
            f"Submit your Borda-count vote by calling submit_vote.  "
            f"You may not vote for yourself.  "
            f"Your vote must rank exactly 5 candidates (scores 5,4,3,2,1) "
            f"and flag exactly 1 candidate as bottom."
        )
        fresh_proxy.initiate_chat(
            recipient=fresh_voter,
            message=dispatch_msg,
            max_turns=vote_max_turns,
        )

    # Dispatch all 21 voters concurrently
    with concurrent.futures.ThreadPoolExecutor(max_workers=max_workers) as executor:
        futures: Dict[concurrent.futures.Future, str] = {
            executor.submit(_run_voter, voter): voter
            for voter in all_pc_slugs
        }
        for future in concurrent.futures.as_completed(futures):
            voter_done: str = futures[future]
            try:
                future.result()
                logger.debug("Phase E: voter '%s' completed.", voter_done)
            except Exception as exc:
                raise RuntimeError(
                    f"Phase E: voter '{voter_done}' failed: {exc}"
                ) from exc

    # Schema validation gate: all 21 vote artifacts must be present and valid
    logger.info("Phase E: Applying fail-closed schema validation gate.")
    failed_voters: List[str] = []
    for slug in all_pc_slugs:
        vote_path: Path = artifact_dir / "voting" / f"vote_{slug}.json"
        try:
            validate_artifact_fn(vote_path, "vote")
        except Exception as exc:
            failed_voters.append(f"{slug}:{exc}")

    if failed_voters:
        raise SchemaValidationError(
            f"Phase E: {len(failed_voters)} vote artifact(s) failed validation "
            f"(fail-closed).\nFailed: {failed_voters}"
        )

    # Load all 21 vote JSON files into the corpus dict keyed by voter slug
    vote_corpus: Dict[str, Any] = {}
    for slug in all_pc_slugs:
        vote_path = artifact_dir / "voting" / f"vote_{slug}.json"
        with open(vote_path, "r", encoding="utf-8") as _vf:
            vote_corpus[slug] = json.load(_vf)

    # Assert the corpus has exactly 21 entries before aggregation
    assert len(vote_corpus) == 21, (
        f"Phase E: vote_corpus has {len(vote_corpus)} entries; expected 21."
    )

    # Persist the full vote corpus for audit and downstream reference
    write_vote_corpus_json(vote_corpus, artifact_dir / "voting")

    # Compute Borda scores via the deterministic scripted aggregation
    borda_scores: Dict[str, float] = compute_borda_aggregation(
        vote_corpus, all_pc_slugs
    )

    # Extract backtest Sharpe ratio as the primary metric score from CRO reports
    # (the composite ranking uses Sharpe as the principal quantitative anchor)
    metric_scores: Dict[str, float] = {}
    for slug in all_pc_slugs:
        cro_path: Path = artifact_dir / "cro_reports" / slug / "cro_report.json"
        with open(cro_path, "r", encoding="utf-8") as _cf:
            cro_data: Dict[str, Any] = json.load(_cf)
        # Extract the backtest Sharpe ratio from the metrics sub-dict
        metric_scores[slug] = float(cro_data["metrics"]["backtest_sharpe"])

    # Compute composite ranking: blends normalised Borda and metric scores
    # alpha controls the regime-dependent weight on Borda vs quantitative metrics
    ranking_result: Dict[str, Any] = compute_composite_ranking(
        borda_scores, metric_scores, alpha, regime
    )

    # Persist the composite ranking for Phase F and Phase G consumption
    composite_ranking_path: Path = artifact_dir / "voting" / "composite_ranking.json"
    with open(composite_ranking_path, "w", encoding="utf-8") as _rf:
        json.dump(ranking_result, _rf, indent=2)

    logger.info("Phase E: Composite ranking generated and persisted.")
    return ranking_result


# -------------------------------------------------------------------------------------------------------------------------------
# Task 19, Step 8 — Execute Phase F (Top-5 Revision)
# -------------------------------------------------------------------------------------------------------------------------------

def execute_phase_f_top5_revision(
    rev_agent:            autogen.ConversableAgent,
    user_proxy:           autogen.UserProxyAgent,
    ranking_result:       Dict[str, Any],
    all_pc_slugs:         List[str],
    pc_registry:          Dict[str, Any],
    artifact_dir:         Path,
    validate_artifact_fn: Callable[[Path, str], bool],
    config:               Dict[str, Any],
    agent_factories:      Dict[str, Any],
) -> List[str]:
    """
    Executes Phase F: applies the diversity constraint to derive the top-5
    shortlist, dispatches fresh revision agents concurrently, and validates
    all produced artifacts.

    Returns the list of top-5 method slugs for consumption by the
    execute_phase_f_cro_rerun callable.

    Corrections versus prior draft
    --------------------------------
    * Diversity constraint assertion enforced post-apply_diversity_constraint.
    * top5_revision directory created with mkdir before write.
    * Fresh revision agent per method via agent_factories["revision"].
    * Fresh UserProxyAgent per thread.
    * Method-specific mu vector injected via factory (not zeros).
    * max_turns sourced from REACT_LOOP_PARAMS.
    * Returns top5_slugs for CRO re-run trigger in orchestrator.
    * Rationale-before-revision requirement injected in dispatch message.
    * Validates both revised weights (if produced) and no_revision flag (if not).

    Parameters
    ----------
    rev_agent : autogen.ConversableAgent
        Template revision agent — NOT used for execution.
    user_proxy : autogen.UserProxyAgent
        Template proxy — NOT used for execution.
    ranking_result : Dict[str, Any]
        Composite ranking dictionary from Phase E.
    all_pc_slugs : List[str]
        Full list of 21 candidate slugs (for family map construction).
    pc_registry : Dict[str, Any]
        PC method registry containing family labels.
    artifact_dir : Path
        Root artifact directory.
    validate_artifact_fn : Callable[[Path, str], bool]
        Bound schema-validation callable.
    config : Dict[str, Any]
        STUDY_CONFIG dictionary.
    agent_factories : Dict[str, Any]
        Factory callables; "revision" key required.

    Returns
    -------
    List[str]
        The top-5 method slugs after diversity enforcement.

    Raises
    ------
    RuntimeError
        If revision execution fails or artifact contracts are not met.
    SchemaValidationError
        If any revised weights artifact fails validation.
    AssertionError
        If the diversity constraint is not satisfied.
    """
    logger.info("Phase F: Applying diversity constraint to derive top-5 shortlist.")

    # Build the family mapping for all 21 candidates
    family_map: Dict[str, str] = {
        s: pc_registry[s]["family"] for s in all_pc_slugs
    }
    # Apply the greedy substitution algorithm to enforce ≥ 3 distinct families
    top5_result: Dict[str, Any] = apply_diversity_constraint(
        ranking_result["composite_ranking"], family_map
    )
    # Extract the final shortlisted method slugs
    top5_slugs: List[str] = top5_result["top_5_methods"]

    # Assert diversity guarantee: at least 3 distinct families in the top-5
    top5_families: Set[str] = {family_map[s] for s in top5_slugs}
    assert len(top5_families) >= 3, (
        f"Phase F diversity assertion failed: top-5 contains only "
        f"{len(top5_families)} distinct families ({top5_families}); "
        f"minimum 3 required.  apply_diversity_constraint may have a defect."
    )

    # Ensure the top5_revision directory exists before writing the selection artifact
    top5_revision_dir: Path = artifact_dir / "top5_revision"
    top5_revision_dir.mkdir(parents=True, exist_ok=True)

    # Persist the top-5 selection with diversity metadata for audit
    top5_selection_path: Path = top5_revision_dir / "top5_selection.json"
    with open(top5_selection_path, "w", encoding="utf-8") as _sf:
        json.dump(top5_result, _sf, indent=2)

    # Extract max_turns for the revision agent from REACT_LOOP_PARAMS
    react_params:    Dict[str, Any] = config["INFRASTRUCTURE"]["REACT_LOOP_PARAMS"]
    rev_max_turns:   int = react_params.get("top5_revision_agent", {}).get(
        "max_iterations", 10
    )

    def _run_reviser(slug: str) -> None:
        """Inner function per revision thread."""
        # Create a fresh revision agent with mu correctly injected for this method
        fresh_rev: autogen.ConversableAgent = agent_factories["revision"](slug)
        # Create a fresh proxy for this revision thread
        fresh_proxy = autogen.UserProxyAgent(
            name=f"proxy_rev_{slug}",
            human_input_mode="NEVER",
            code_execution_config=False,
        )
        dispatch_msg: str = (
            f"You are '{slug}'.  Review your peer critiques in "
            f"{str(artifact_dir / 'peer_reviews')} and your CRO report at "
            f"{str(artifact_dir / 'cro_reports' / slug / 'cro_report.json')}.  "
            f"Before calling run_optimizer, you must explicitly state in your "
            f"reasoning which peer criticisms you accept and which you reject, "
            f"and why.  This rationale is required before any tool call.  "
            f"If your analysis supports revision, call run_optimizer with "
            f"updated constraints, then check_ips_compliance, then "
            f"write_pc_weights_json, then write_pc_report_md.  "
            f"If no revision is warranted, call confirm_no_revision immediately."
        )
        fresh_proxy.initiate_chat(
            recipient=fresh_rev,
            message=dispatch_msg,
            max_turns=rev_max_turns,
        )

    # Dispatch the 5 revision agents concurrently
    with concurrent.futures.ThreadPoolExecutor(max_workers=5) as executor:
        futures: Dict[concurrent.futures.Future, str] = {
            executor.submit(_run_reviser, slug): slug
            for slug in top5_slugs
        }
        for future in concurrent.futures.as_completed(futures):
            slug_done: str = futures[future]
            try:
                future.result()
                logger.debug("Phase F: reviser '%s' completed.", slug_done)
            except Exception as exc:
                raise RuntimeError(
                    f"Phase F: reviser '{slug_done}' execution failed: {exc}"
                ) from exc

    # Validate Phase F artifacts: each method must produce either revised weights
    # or a no_revision confirmation — no other outcome is acceptable
    logger.info("Phase F: Applying fail-closed artifact validation gate.")
    for slug in top5_slugs:
        # Check primary path: revised weights JSON
        revised_path: Path = (
            artifact_dir / "top5_revision" / slug / "revised_pc_weights.json"
        )
        # Check alternative path: no-revision confirmation JSON
        no_rev_path: Path = (
            artifact_dir / "top5_revision" / slug / "no_revision_confirmed.json"
        )
        if revised_path.exists():
            # Revised weights exist — validate against schema
            try:
                validate_artifact_fn(revised_path, "pc_weights")
            except Exception as exc:
                raise SchemaValidationError(
                    f"Phase F: revised_pc_weights.json for '{slug}' failed "
                    f"validation: {exc}"
                ) from exc
        elif no_rev_path.exists():
            # No revision confirmed — acceptable terminal state for this agent
            logger.debug(
                "Phase F: '%s' confirmed no revision — no_revision_confirmed.json present.",
                slug,
            )
        else:
            # Neither artifact found — pipeline halted (fail-closed)
            raise RuntimeError(
                f"Phase F: '{slug}' produced neither revised_pc_weights.json "
                f"nor no_revision_confirmed.json.  Pipeline halted (fail-closed)."
            )

    logger.info(
        "Phase F: Top-5 revision complete.  Shortlist: %s.", top5_slugs
    )
    # Return top5_slugs for CRO re-run targeting in the orchestrator
    return top5_slugs


# -------------------------------------------------------------------------------------------------------------------------------
# Task 19, Step 7B — Execute Phase F CRO Re-Run (Post-Revision)
# -------------------------------------------------------------------------------------------------------------------------------

def execute_phase_f_cro_rerun(
    top5_slugs:           List[str],
    artifact_dir:         Path,
    validate_artifact_fn: Callable[[Path, str], bool],
    config:               Dict[str, Any],
    agent_factories:      Dict[str, Any],
    max_workers:          int = 5,
) -> None:
    """
    Re-runs the CRO risk report script tools for every top-5 method that
    produced revised weights in Phase F.  Overwrites the stale cro_report.json
    with freshly computed metrics before Phase G begins.

    Per Task 30, Step 3: "For any agent that altered its weights, the
    orchestrator must automatically re-trigger Task 27 (CRO metrics) to
    generate an updated cro_report.json."  Omitting this step would cause
    the CIO agent to evaluate revised portfolios using stale risk metrics.

    Parameters
    ----------
    top5_slugs : List[str]
        The top-5 method slugs after diversity enforcement.
    artifact_dir : Path
        Root artifact directory.
    validate_artifact_fn : Callable[[Path, str], bool]
        Bound schema-validation callable.
    config : Dict[str, Any]
        STUDY_CONFIG dictionary.
    agent_factories : Dict[str, Any]
        Factory callables; "cro" key required.
    max_workers : int, optional
        Maximum concurrent threads.  Defaults to 5 (one per top-5 method).

    Raises
    ------
    RuntimeError
        If any CRO re-run fails.
    SchemaValidationError
        If any updated cro_report.json fails schema validation.
    """
    logger.info(
        "Phase F CRO Re-run: Identifying top-5 methods with revised weights."
    )

    # Identify which top-5 methods actually revised their weights
    revised_slugs: List[str] = [
        slug for slug in top5_slugs
        if (artifact_dir / "top5_revision" / slug / "revised_pc_weights.json").exists()
    ]
    if not revised_slugs:
        # No revisions were made; no re-run is necessary
        logger.info("Phase F CRO Re-run: No revisions detected — skipping re-run.")
        return

    logger.info(
        "Phase F CRO Re-run: Triggering CRO re-run for revised methods: %s.",
        revised_slugs,
    )

    # Extract CRO max_turns from REACT_LOOP_PARAMS
    react_params:  Dict[str, Any] = config["INFRASTRUCTURE"]["REACT_LOOP_PARAMS"]
    cro_max_turns: int = react_params.get("cro_narrative_agent", {}).get(
        "max_iterations", 20
    )
    # Mandatory CRO tool-call sequence for the re-run (same as Phase C)
    cro_tool_sequence: List[str] = react_params.get(
        "cro_narrative_agent", {}
    ).get("mandatory_tool_sequence", [])

    def _rerun_cro_for_revised(slug: str) -> None:
        """Inner function per CRO re-run thread."""
        # Create a fresh CRO agent configured for the revised weights path
        fresh_cro: autogen.ConversableAgent = agent_factories["cro"](slug)
        fresh_proxy = autogen.UserProxyAgent(
            name=f"proxy_cro_rerun_{slug}",
            human_input_mode="NEVER",
            code_execution_config=False,
        )
        # Point the CRO re-run at the REVISED weights (not the original weights)
        revised_weights_path: Path = (
            artifact_dir / "top5_revision" / slug / "revised_pc_weights.json"
        )
        dispatch_msg: str = (
            f"Generate an UPDATED CRO risk report for candidate '{slug}' "
            f"using the REVISED weights at: {str(revised_weights_path)}.  "
            f"Overwrite the previous cro_report.json in "
            f"{str(artifact_dir / 'cro_reports' / slug)}.  "
            f"Call tools in this order: {cro_tool_sequence}."
        )
        fresh_proxy.initiate_chat(
            recipient=fresh_cro,
            message=dispatch_msg,
            max_turns=cro_max_turns,
        )

    # Dispatch CRO re-runs for all revised methods concurrently
    with concurrent.futures.ThreadPoolExecutor(max_workers=max_workers) as executor:
        futures: Dict[concurrent.futures.Future, str] = {
            executor.submit(_rerun_cro_for_revised, slug): slug
            for slug in revised_slugs
        }
        for future in concurrent.futures.as_completed(futures):
            slug_done: str = futures[future]
            try:
                future.result()
                logger.debug(
                    "Phase F CRO Re-run: '%s' updated CRO report completed.",
                    slug_done,
                )
            except Exception as exc:
                raise RuntimeError(
                    f"Phase F CRO Re-run: execution failed for '{slug_done}': {exc}"
                ) from exc

    # Validate all updated CRO reports (fail-closed)
    logger.info("Phase F CRO Re-run: Validating updated CRO report artifacts.")
    for slug in revised_slugs:
        updated_cro_path: Path = (
            artifact_dir / "cro_reports" / slug / "cro_report.json"
        )
        try:
            validate_artifact_fn(updated_cro_path, "cro_report")
        except Exception as exc:
            raise SchemaValidationError(
                f"Phase F CRO Re-run: updated cro_report.json for '{slug}' "
                f"failed validation: {exc}"
            ) from exc

    logger.info(
        "Phase F CRO Re-run: %d updated CRO report(s) validated successfully.",
        len(revised_slugs),
    )


# -------------------------------------------------------------------------------------------------------------------------------
# Task 19, Step 8 — Execute Phase G (CIO Selection)
# -------------------------------------------------------------------------------------------------------------------------------

def execute_phase_g_cio_selection(
    cio_agent:            autogen.ConversableAgent,
    user_proxy:           autogen.UserProxyAgent,
    artifact_dir:         Path,
    validate_artifact_fn: Callable[[Path, str], bool],
    config:               Dict[str, Any],
    top5_slugs:           List[str],
    sigma:                np.ndarray,
    benchmark_weights:    np.ndarray,
    ips_constraints:      Dict[str, Any],
) -> Tuple[Path, Path]:
    """
    Executes Phase G: enforces the pre-flight artifact gate, dispatches the
    CIO agent with xhigh reasoning effort, applies the mandatory tool-call
    sequence, performs post-hoc IPS compliance verification on the final
    weights, and validates all output artifacts.

    Corrections versus prior draft
    --------------------------------
    * Pre-flight artifact gate verifies all upstream artifacts before CIO
      agent is activated.
    * max_turns sourced from REACT_LOOP_PARAMS (max_iterations for cio_agent).
    * Mandatory tool-call sequence injected into dispatch message.
    * Post-hoc IPS arithmetic check on final_weights.json (dual-layer gate).
    * Existence assertion on final_weights.json before returning.
    * CIO composite score weight-sum assertion (0.25+0.15+0.15+0.20+0.15+0.10=1.0).
    * SchemaValidationError raised (defined in this module).
    * Fresh UserProxyAgent for this phase.

    Parameters
    ----------
    cio_agent : autogen.ConversableAgent
        The instantiated CIO agent with all tools registered.
    user_proxy : autogen.UserProxyAgent
        Template proxy — NOT used for execution; fresh proxy created internally.
    artifact_dir : Path
        Root artifact directory.
    validate_artifact_fn : Callable[[Path, str], bool]
        Bound schema-validation callable.
    config : Dict[str, Any]
        STUDY_CONFIG dictionary.
    top5_slugs : List[str]
        Top-5 method slugs for pre-flight artifact verification.
    sigma : np.ndarray
        18×18 annualized covariance matrix for post-hoc IPS check.
    benchmark_weights : np.ndarray
        18-element benchmark weight vector for post-hoc TE check.
    ips_constraints : Dict[str, Any]
        IPS constraint parameters for post-hoc compliance check.

    Returns
    -------
    Tuple[Path, Path]
        Paths to final_weights.json and board_memo.md.

    Raises
    ------
    RuntimeError
        If the pre-flight gate fails, execution fails, post-hoc IPS check
        fails, or the board memo is absent.
    SchemaValidationError
        If final_weights.json fails schema validation.
    """
    logger.info("Phase G: Executing CIO pre-flight artifact gate.")

    # ---- Pre-flight artifact gate ----
    # Verify composite ranking exists
    composite_ranking_path: Path = artifact_dir / "voting" / "composite_ranking.json"
    if not composite_ranking_path.exists():
        raise RuntimeError(
            "Phase G pre-flight gate failed: composite_ranking.json missing.  "
            "Phase E must complete before Phase G."
        )
    # Verify vote corpus exists
    vote_corpus_path: Path = artifact_dir / "voting" / "vote_corpus.json"
    if not vote_corpus_path.exists():
        raise RuntimeError(
            "Phase G pre-flight gate failed: vote_corpus.json missing."
        )
    # Verify each top-5 method has a final weights artifact (revised or original)
    for slug in top5_slugs:
        # Accept revised weights if they exist; fall back to original Phase A weights
        revised_path: Path = (
            artifact_dir / "top5_revision" / slug / "revised_pc_weights.json"
        )
        original_path: Path = artifact_dir / "pc_agents" / slug / "pc_weights.json"
        confirmed_no_rev: Path = (
            artifact_dir / "top5_revision" / slug / "no_revision_confirmed.json"
        )
        if not (revised_path.exists() or original_path.exists() or confirmed_no_rev.exists()):
            raise RuntimeError(
                f"Phase G pre-flight gate failed: no final weights or revision "
                f"confirmation found for top-5 candidate '{slug}'."
            )
    logger.info("Phase G: Pre-flight artifact gate passed.")

    # ---- CIO composite score weight-sum assertion ----
    # The formula is CIO_Score = 0.25*SR + 0.15*IPS + 0.15*Div + 0.20*RF + 0.15*ER + 0.10*CU
    cio_weights: List[float] = [0.25, 0.15, 0.15, 0.20, 0.15, 0.10]
    assert abs(sum(cio_weights) - 1.0) < 1e-9, (
        f"CIO composite score weights do not sum to 1.0: {sum(cio_weights)}"
    )

    # Extract CIO max_turns from REACT_LOOP_PARAMS
    react_params: Dict[str, Any] = config["INFRASTRUCTURE"]["REACT_LOOP_PARAMS"]
    cio_max_turns: int = react_params.get("cio_agent", {}).get("max_iterations", 25)
    # Extract the mandatory tool-call sequence for the CIO agent
    cio_sequence: List[str] = react_params.get("cio_agent", {}).get(
        "mandatory_tool_sequence", []
    )

    # Create a fresh UserProxyAgent for Phase G to prevent state contamination
    fresh_proxy_g = autogen.UserProxyAgent(
        name="proxy_cio",
        human_input_mode="NEVER",
        code_execution_config=False,
    )

    dispatch_msg: str = (
        "Execute the CIO selection protocol.  "
        f"You must call tools in this mandatory order: {cio_sequence}.  "
        f"All upstream artifacts are confirmed present at {str(artifact_dir)}.  "
        "Evaluate all seven ensemble methods, select the highest-scoring "
        "IPS-compliant ensemble, and generate the Board Memo.  "
        "Use xhigh reasoning effort."
    )

    try:
        fresh_proxy_g.initiate_chat(
            recipient=cio_agent,
            message=dispatch_msg,
            max_turns=cio_max_turns,
        )
    except Exception as exc:
        raise RuntimeError(
            f"Phase G: CIO agent execution failed: {exc}"
        ) from exc

    # ---- Schema validation gate ----
    logger.info("Phase G: Applying schema validation gate on final_weights.json.")
    final_weights_path: Path = artifact_dir / "cio" / "final_weights.json"
    try:
        validate_artifact_fn(final_weights_path, "final_weights")
    except Exception as exc:
        raise SchemaValidationError(
            f"Phase G: final_weights.json failed schema validation: {exc}"
        ) from exc

    # ---- Existence assertion on final_weights.json ----
    if not final_weights_path.exists():
        raise RuntimeError(
            "Phase G: final_weights.json does not exist after schema validation — "
            "this indicates a validation function defect."
        )

    # ---- Post-hoc IPS arithmetic check (dual-layer gate) ----
    # Load the final weights for independent arithmetic verification
    with open(final_weights_path, "r", encoding="utf-8") as _fw:
        final_weights_data: Dict[str, Any] = json.load(_fw)
    final_weights_arr: np.ndarray = np.array(
        final_weights_data["weights"], dtype=np.float64
    )
    # Check 1: weights must be non-negative (long-only constraint)
    if np.any(final_weights_arr < -1e-9):
        raise RuntimeError(
            f"Phase G post-hoc IPS check failed: final weights contain negative "
            f"values (long-only violation).  Min weight: {final_weights_arr.min():.6f}"
        )
    # Check 2: weights must sum to 1.0 (budget constraint)
    weight_sum: float = float(final_weights_arr.sum())
    if abs(weight_sum - 1.0) > 1e-6:
        raise RuntimeError(
            f"Phase G post-hoc IPS check failed: final weights sum to "
            f"{weight_sum:.8f}, not 1.0."
        )
    # Check 3: no individual weight may exceed the IPS maximum (25%)
    max_weight: float = float(final_weights_arr.max())
    ips_max:    float = float(ips_constraints.get("max_weight_per_asset", 0.25))
    if max_weight > ips_max + 1e-9:
        raise RuntimeError(
            f"Phase G post-hoc IPS check failed: maximum weight {max_weight:.4f} "
            f"exceeds IPS bound {ips_max:.4f}."
        )
    # Check 4: ex-ante portfolio volatility must fall within the IPS band (8–12%)
    vol_target:   Tuple[float, float] = tuple(
        config["IPS_GOVERNANCE"]["OBJECTIVE"]["VOLATILITY_BAND"]
    )
    ex_ante_vol:  float = float(
        np.sqrt(final_weights_arr @ sigma @ final_weights_arr)
    )
    if not (vol_target[0] - 1e-4 <= ex_ante_vol <= vol_target[1] + 1e-4):
        raise RuntimeError(
            f"Phase G post-hoc IPS check failed: ex-ante portfolio volatility "
            f"{ex_ante_vol:.4f} is outside the IPS band "
            f"[{vol_target[0]:.2f}, {vol_target[1]:.2f}]."
        )
    # Check 5: tracking error must not exceed the 6% TE budget
    te_budget: float = float(
        config["IPS_GOVERNANCE"]["ACTIVE_RISK"]["TRACKING_ERROR_BUDGET_EX_ANTE"]
    )
    active_weights: np.ndarray = final_weights_arr - benchmark_weights
    tracking_error: float = float(
        np.sqrt(active_weights @ sigma @ active_weights)
    )
    if tracking_error > te_budget + 1e-9:
        raise RuntimeError(
            f"Phase G post-hoc IPS check failed: tracking error {tracking_error:.4f} "
            f"exceeds the TE budget {te_budget:.4f}."
        )
    logger.info(
        "Phase G post-hoc IPS check passed (vol=%.4f, TE=%.4f, max_w=%.4f).",
        ex_ante_vol, tracking_error, max_weight,
    )

    # ---- Board memo existence check ----
    board_memo_path: Path = artifact_dir / "cio" / "board_memo.md"
    if not board_memo_path.exists():
        raise RuntimeError(
            "Phase G: board_memo.md was not generated by the CIO agent."
        )
    # Minimum content check: board memo must be non-trivial in length
    board_memo_size: int = board_memo_path.stat().st_size
    if board_memo_size < 500:
        raise RuntimeError(
            f"Phase G: board_memo.md is suspiciously small ({board_memo_size} bytes) — "
            f"the CIO agent may have produced a truncated response."
        )

    logger.info("Phase G: CIO selection completed successfully.")
    return final_weights_path, board_memo_path


# -------------------------------------------------------------------------------------------------------------------------------
# Task 19, Orchestrator — Full Strategy Review Pipeline
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_19_strategy_review(
    config:               Dict[str, Any],
    sigma:                np.ndarray,
    mu_dict:              Dict[str, np.ndarray],
    rf:                   float,
    benchmark_weights:    np.ndarray,
    returns_matrix:       np.ndarray,
    returns_index:        pd.DatetimeIndex,
    factors_df:           pd.DataFrame,
    macro_view_path:      Path,
    cma_dir:              Path,
    run_output_dir:       Path,
    openai_api_key:       str,
    validate_artifact_fn: Callable[[Path, str], bool],
) -> Tuple[Path, Path]:
    """
    Orchestrates the full Strategy Review pipeline (Task 19). Instantiates the
    AutoGen MAS topology, executes the seven sequential phases via the FSM-
    controlled StrategyReviewController, triggers the Phase F CRO re-run for
    any revised portfolios, and returns the final allocation artifacts.

    CRITICAL REMEDIATION: Resolves the CIO Temporal Paradox by explicitly
    injecting the `top5_slugs` generated in Phase F into the `pipeline_state`
    dictionary of the MASTopology. This ensures the CIO Agent's tools have
    access to the mathematically required shortlist during Phase G execution.

    Parameters
    ----------
    config : Dict[str, Any]
        The validated STUDY_CONFIG dictionary.
    sigma : np.ndarray
        18x18 annualized shrinkage covariance matrix (from Task 24).
    mu_dict : Dict[str, np.ndarray]
        Method-slug -> 18-element expected-return vector mapping.
    rf : float
        Annualized risk-free rate.
    benchmark_weights : np.ndarray
        18-element benchmark weight vector.
    returns_matrix : np.ndarray
        T x 18 monthly returns matrix.
    returns_index : pd.DatetimeIndex
        The temporal index corresponding to the returns_matrix.
    factors_df : pd.DataFrame
        Monthly Fama-French factor returns dataframe.
    macro_view_path : Path
        Path to the validated macro-view.json artifact.
    cma_dir : Path
        Directory containing per-asset cma.json artifacts.
    run_output_dir : Path
        Root directory for all run artifacts.
    openai_api_key : str
        OpenAI API key.
    validate_artifact_fn : Callable[[Path, str], bool]
        Bound schema-validation callable from Task 16.

    Returns
    -------
    Tuple[Path, Path]
        Paths to final_weights.json and board_memo.md.

    Raises
    ------
    RuntimeError
        If any phase fails. A failure manifest is written before re-raising.
    """
    # Log the commencement of the Strategy Review pipeline orchestration
    logger.info("Task 19: Starting Strategy Review pipeline orchestration.")

    # -----------------------------------------------------------------------
    # Record configuration SHA-256 hash in run manifest
    # -----------------------------------------------------------------------
    # Serialize config to a canonical JSON string for hashing
    config_json_bytes: bytes = json.dumps(config, sort_keys=True, default=str).encode()
    config_hash: str = hashlib.sha256(config_json_bytes).hexdigest()
    pipeline_start_utc: str = datetime.now(tz=timezone.utc).isoformat()
    run_manifest_path: Path = run_output_dir / "run_manifest.json"

    # Merge with any existing manifest entries to preserve prior task records
    run_manifest: Dict[str, Any] = {}
    if run_manifest_path.exists():
        with open(run_manifest_path, "r", encoding="utf-8") as _rm:
            run_manifest = json.load(_rm)

    run_manifest["task_19"] = {
        "pipeline_start_utc": pipeline_start_utc,
        "config_sha256": config_hash,
        "status": "RUNNING",
    }

    run_manifest_path.parent.mkdir(parents=True, exist_ok=True)
    with open(run_manifest_path, "w", encoding="utf-8") as _rm:
        json.dump(run_manifest, _rm, indent=2)

    # -----------------------------------------------------------------------
    # Load macro view once for regime and alpha extraction
    # -----------------------------------------------------------------------
    with open(macro_view_path, "r", encoding="utf-8") as _mf:
        macro_view_data: Dict[str, Any] = json.load(_mf)

    # Extract current regime from the macro view for alpha schedule lookup
    regime: str = macro_view_data.get("regime", "Expansion")

    # Extract alpha from config
    # alpha_schedule must be defined under METHODOLOGY_PARAMS in STUDY_CONFIG
    alpha_schedule: Dict[str, float] = config.get(
        "METHODOLOGY_PARAMS", {}
    ).get(
        "STRATEGY_REVIEW", {}
    ).get(
        "vote_metric_blend_alpha_schedule",
        # Fallback to a sensible default if not yet specified in config
        {"Expansion": 0.5, "Late-cycle": 0.6, "Recession": 0.4, "Recovery": 0.5},
    )
    alpha: float = float(alpha_schedule.get(regime, 0.5))
    logger.info("Orchestrator: regime='%s', alpha=%.2f from config.", regime, alpha)

    # -----------------------------------------------------------------------
    # Extract per-phase max_workers from config
    # -----------------------------------------------------------------------
    parallel_cfg: Dict[str, int] = config["INFRASTRUCTURE"]["AUTOGEN_TOPOLOGY"]["parallel_dispatch"]
    max_workers_a: int = parallel_cfg.get("max_workers_pc_submission", 8)
    max_workers_c: int = parallel_cfg.get("max_workers_cro_reports", 8)
    max_workers_d: int = parallel_cfg.get("max_workers_peer_review", 8)
    max_workers_e: int = parallel_cfg.get("max_workers_voting", 8)

    # -----------------------------------------------------------------------
    # Construct the canonical all_pc_slugs list
    # -----------------------------------------------------------------------
    # Derived from pc_reg keys excluding non-PC entries; adversarial appended last
    pc_reg: Dict[str, Any] = config["REGISTRIES"]["PC_AGENTS"]["method_configs"]
    all_pc_slugs: List[str] = [
        s for s in pc_reg.keys()
        if s not in {"researcher", "adversarial_diversifier"}
    ]
    all_pc_slugs.append("adversarial_diversifier")

    # Assert the exact number of PC slugs
    assert len(all_pc_slugs) == 21, (
        f"Orchestrator: expected 21 PC slugs, derived {len(all_pc_slugs)}. "
        f"Check pc_registry entries."
    )

    # -----------------------------------------------------------------------
    # Instantiate the full MAS topology
    # -----------------------------------------------------------------------
    topology: MASTopology = instantiate_mas_topology(
        config=config,
        sigma=sigma,
        mu_dict=mu_dict,
        rf=rf,
        benchmark_weights=benchmark_weights,
        returns_matrix=returns_matrix,
        returns_index=returns_index,
        factors_df=factors_df,
        macro_view_path=macro_view_path,
        cma_dir=cma_dir,
        artifact_dir=run_output_dir,
        openai_api_key=openai_api_key,
        validate_artifact_fn=validate_artifact_fn,
    )

    # Template proxy used only to satisfy function signatures; each phase creates
    # its own fresh UserProxyAgent internally
    _template_proxy = autogen.UserProxyAgent(
        name="System_Orchestrator",
        human_input_mode="NEVER",
        code_execution_config=False,
    )

    # Pipeline-wide try/except: guarantees a failure manifest is written on any error
    try:

        # ================================================================
        # PHASE A: PC Weight Submission
        # ================================================================
        logger.info("Orchestrator: Executing Phase A.")
        execute_phase_a_pc_submission(
            phase_a_agents=topology.agents["Phase_A"],
            user_proxy=_template_proxy,
            artifact_dir=run_output_dir,
            validate_artifact_fn=validate_artifact_fn,
            config=config,
            openai_api_key=openai_api_key,
            agent_factories=topology.agent_factories,
            max_workers=max_workers_a,
        )

        # ================================================================
        # PHASE B: Adversarial Diversifier
        # ================================================================
        logger.info("Orchestrator: Executing Phase B.")
        execute_phase_b_adversarial_diversifier(
            adv_agent=topology.agents["Phase_B"]["adversarial_diversifier"],
            user_proxy=_template_proxy,
            artifact_dir=run_output_dir,
            validate_artifact_fn=validate_artifact_fn,
            config=config,
            agent_factories=topology.agent_factories,
        )

        # ================================================================
        # PHASE C: CRO Risk Reports
        # ================================================================
        logger.info("Orchestrator: Executing Phase C.")
        execute_phase_c_cro_reports(
            cro_agent=topology.agents["Phase_C"]["cro_agent"],
            user_proxy=_template_proxy,
            all_pc_slugs=all_pc_slugs,
            artifact_dir=run_output_dir,
            validate_artifact_fn=validate_artifact_fn,
            config=config,
            agent_factories=topology.agent_factories,
            max_workers=max_workers_c,
        )

        # ================================================================
        # PHASE D: Peer Review
        # ================================================================
        logger.info("Orchestrator: Executing Phase D.")
        execute_phase_d_peer_review(
            peer_agent=topology.agents["Phase_D"]["peer_review_agent"],
            user_proxy=_template_proxy,
            all_pc_slugs=all_pc_slugs,
            pc_registry=topology.pc_registry,
            seed=config["RANDOM_SEEDS"]["peer_review_assignment_seed"],
            artifact_dir=run_output_dir,
            config=config,
            agent_factories=topology.agent_factories,
            max_workers=max_workers_d,
        )

        # ================================================================
        # PHASE E: Voting and Aggregation
        # ================================================================
        logger.info("Orchestrator: Executing Phase E.")

        # 1. Execute Voting Dispatch
        vote_corpus: Dict[str, Dict[str, Any]] = execute_phase_e_voting_dispatch(
            all_pc_slugs=all_pc_slugs,
            artifact_dir=run_output_dir,
            validate_artifact_fn=validate_artifact_fn,
            config=config,
            agent_factories=topology.agent_factories,
            max_workers=max_workers_e
        )

        # 2. Persist the corpus
        write_vote_corpus_json(vote_corpus, run_output_dir / "voting")

        # 3. Aggregate Borda Scores
        borda_scores: Dict[str, float] = compute_borda_aggregation(vote_corpus, all_pc_slugs)

        # 4. Extract Metric Scores (Backtest Sharpe) from CRO reports
        metric_scores: Dict[str, float] = {}
        for slug in all_pc_slugs:
            cro_path: Path = run_output_dir / "cro_reports" / slug / "cro_report.json"
            if not cro_path.exists():
                raise FileNotFoundError(f"CRO report missing for {slug} at {cro_path}")
            with open(cro_path, "r", encoding="utf-8") as f:
                cro_data: Dict[str, Any] = json.load(f)
            metric_scores[slug] = float(cro_data["metrics"]["backtest_sharpe"])

        # 5. Compute Composite Ranking
        ranking_result: Dict[str, Any] = compute_composite_ranking(
            borda_scores=borda_scores,
            metric_scores=metric_scores,
            alpha=alpha,
            regime=regime
        )

        # Persist the composite ranking
        ranking_path: Path = run_output_dir / "voting" / "composite_ranking.json"
        try:
            with open(ranking_path, "w", encoding="utf-8") as f:
                json.dump(ranking_result, f, indent=4)
        except OSError as exc:
            raise OSError(f"Failed to write composite_ranking.json: {exc}") from exc

        # ================================================================
        # PHASE F: Top-5 Revision
        # ================================================================
        logger.info("Orchestrator: Executing Phase F.")
        top5_slugs: List[str] = orchestrate_task_30_top5_revision(
            ranking_result=ranking_result,
            all_pc_slugs=all_pc_slugs,
            pc_registry=topology.pc_registry,
            config=config,
            run_output_dir=run_output_dir,
            validate_artifact_fn=validate_artifact_fn,
            agent_factories=topology.agent_factories,
            sigma=sigma,
            mu_dict=mu_dict,
            benchmark_weights=benchmark_weights,
            returns_matrix=returns_matrix,
            returns_index=returns_index,
            factors_df=factors_df,
            rf=rf
        )

        # ================================================================
        # CRITICAL REMEDIATION: State Injection
        # ================================================================
        # Inject the top5_slugs into the pipeline_state dictionary.
        # Because the CIO agent's tools hold a closure reference to this dictionary,
        # they will now successfully resolve the "top5_slugs" key at runtime during Phase G.
        logger.info("Orchestrator: Injecting top5_slugs into pipeline_state for Phase G consumption.")
        topology.pipeline_state["top5_slugs"] = top5_slugs

        # ================================================================
        # PHASE G: CIO Selection
        # ================================================================
        logger.info("Orchestrator: Executing Phase G.")
        final_weights_path, board_memo_path = execute_phase_g_cio_selection(
            cio_agent=topology.agents["Phase_G"]["cio_agent"],
            user_proxy=_template_proxy,
            artifact_dir=run_output_dir,
            validate_artifact_fn=validate_artifact_fn,
            config=config,
            top5_slugs=top5_slugs,
            sigma=sigma,
            benchmark_weights=benchmark_weights,
            ips_constraints=config["IPS_GOVERNANCE"]["IMPLEMENTATION_CONSTRAINTS"],
        )

    except Exception as exc:
        # ---- Write failure manifest before re-raising ----
        pipeline_fail_utc: str = datetime.now(tz=timezone.utc).isoformat()
        run_manifest["task_19"]["status"] = "FAILED"
        run_manifest["task_19"]["failure_utc"] = pipeline_fail_utc
        run_manifest["task_19"]["failure_message"] = str(exc)

        with open(run_manifest_path, "w", encoding="utf-8") as _rm:
            json.dump(run_manifest, _rm, indent=2)

        logger.critical(
            "Task 19: Pipeline failed. Failure manifest written to %s. Error: %s",
            run_manifest_path, exc,
        )
        raise

    # ---- Update run manifest with success record ----
    pipeline_end_utc: str = datetime.now(tz=timezone.utc).isoformat()
    run_manifest["task_19"]["status"] = "SUCCESS"
    run_manifest["task_19"]["pipeline_end_utc"] = pipeline_end_utc
    run_manifest["task_19"]["final_weights_path"] = str(final_weights_path)
    run_manifest["task_19"]["board_memo_path"] = str(board_memo_path)

    with open(run_manifest_path, "w", encoding="utf-8") as _rm:
        json.dump(run_manifest, _rm, indent=2)

    logger.info(
        "Task 19: Strategy Review pipeline completed successfully. "
        "Final weights: %s. Board memo: %s.",
        final_weights_path, board_memo_path,
    )

    # Return the paths to the terminal artifacts
    return final_weights_path, board_memo_path
````

````python
# Task 20 — Implement Macro scoring and regime classification (weighted scoring framework)

# ==============================================================================
# Task 20: Implement Macro scoring and regime classification
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 20, Step 1: Define transformations from raw macro series to standardized dimension scores
# -------------------------------------------------------------------------------------------------------------------------------

def compute_expanding_zscores(
    df_macro_raw: pd.DataFrame,
    as_of_date: str,
    epsilon: float = 1e-8
) -> pd.DataFrame:
    """
    Computes expanding z-scores for all numeric columns in the macro dataframe
    up to the specified as-of date, strictly preventing lookahead bias.

    CRITICAL REMEDIATION: Replaces the exact equality check `replace(0.0, epsilon)`
    with a robust boolean mask `expanding_std[expanding_std < epsilon] = epsilon`.
    This guarantees numerical stability against floating-point drift (e.g., 1e-16)
    that would otherwise cause z-scores to explode and corrupt the softmax confidence.

    Equation
    --------
    z_t = (x_t - \mu_{1:t}) / \sigma_{1:t}

    Parameters
    ----------
    df_macro_raw : pd.DataFrame
        The raw macroeconomic indicator dataframe. Must have a DatetimeIndex.
    as_of_date : str
        ISO-8601 date string defining the point-in-time boundary.
    epsilon : float, optional
        Small positive constant applied as a floor to \sigma_{1:t} to prevent
        division by zero or near-zero. Defaults to 1e-8.

    Returns
    -------
    pd.DataFrame
        DataFrame of expanding z-scores indexed to the same DatetimeIndex.

    Raises
    ------
    TypeError
        If df_macro_raw is not a pd.DataFrame or its index is not a pd.DatetimeIndex.
    ValueError
        If lookahead contamination is detected, if the dataframe is empty,
        or if epsilon is not strictly positive.
    """
    # Guard: verify the input is a pandas DataFrame
    if not isinstance(df_macro_raw, pd.DataFrame):
        raise TypeError(f"df_macro_raw must be a pd.DataFrame; received {type(df_macro_raw).__name__}.")

    # Guard: verify the index is a DatetimeIndex for temporal comparisons
    if not isinstance(df_macro_raw.index, pd.DatetimeIndex):
        raise TypeError("df_macro_raw.index must be a pd.DatetimeIndex.")

    # Guard: epsilon must be strictly positive to prevent a zero denominator
    if epsilon <= 0.0:
        raise ValueError(f"epsilon must be strictly positive; received {epsilon}.")

    # Parse the as-of date string into a pandas Timestamp
    try:
        as_of_ts: pd.Timestamp = pd.Timestamp(as_of_date)
    except Exception as exc:
        raise ValueError(f"as_of_date '{as_of_date}' cannot be parsed as a Timestamp: {exc}") from exc

    # Create a working copy to avoid mutating the caller's original dataframe
    df_clean: pd.DataFrame = df_macro_raw.copy()

    # Strip timezone from the DatetimeIndex if it is tz-aware
    if df_clean.index.tz is not None:
        df_clean.index = df_clean.index.tz_convert(None)

    # Strip timezone from the as-of Timestamp to match the naive index
    if as_of_ts.tzinfo is not None:
        as_of_ts = as_of_ts.tz_convert(None).tz_localize(None)

    # POINT-IN-TIME ASSERTION (fail-closed)
    # Hard gate to surface upstream data pipeline contamination
    n_future_rows: int = int((df_clean.index > as_of_ts).sum())
    if n_future_rows > 0:
        raise ValueError(
            f"Lookahead contamination: df_macro_raw contains {n_future_rows} row(s) "
            f"with dates strictly beyond AS_OF_DATE='{as_of_date}'."
        )

    # Retain all rows at or before the as-of date
    df_filtered: pd.DataFrame = df_clean.loc[df_clean.index <= as_of_ts]

    # Guard: the filtered frame must be non-empty
    if df_filtered.empty:
        raise ValueError(f"df_macro_raw contains no observations on or before '{as_of_date}'.")

    # Restrict to numeric columns only
    df_numeric: pd.DataFrame = df_filtered.select_dtypes(include=[np.number])

    if df_numeric.empty:
        raise ValueError("df_macro_raw contains no numeric columns eligible for z-score computation.")

    # Compute the expanding mean: \mu_{1:t}
    expanding_mean: pd.DataFrame = df_numeric.expanding(min_periods=1).mean()

    # Compute the expanding standard deviation: \sigma_{1:t} (ddof=1)
    expanding_std: pd.DataFrame = df_numeric.expanding(min_periods=2).std(ddof=1)

    # CRITICAL REMEDIATION: Robust boolean masking for the epsilon floor
    # Fill NaNs (from the first observation) with epsilon
    expanding_std_safe: pd.DataFrame = expanding_std.fillna(epsilon)

    # Apply a strict boolean mask to catch exact zeros AND floating-point drift (e.g., 1e-16)
    # This guarantees the denominator is never smaller than epsilon
    expanding_std_safe[expanding_std_safe < epsilon] = epsilon

    # Compute z_t = (x_t - \mu_{1:t}) / \sigma_{1:t}
    df_zscores: pd.DataFrame = (df_numeric - expanding_mean) / expanding_std_safe

    logger.debug(
        f"compute_expanding_zscores: produced z-scores for {df_zscores.shape[1]} column(s) "
        f"over {df_zscores.shape[0]} observation(s) up to '{as_of_date}'."
    )

    # Return the expanding z-score dataframe
    return df_zscores


def compute_dimension_scores(
    df_zscores:           pd.DataFrame,
    scoring_weights:      Dict[str, Dict[str, float]],
    dimension_key_mapping: Dict[str, str],
) -> Dict[str, float]:
    """
    Computes the four standardized dimension scores (Growth, Inflation, Policy,
    Financial Conditions) by applying the frozen configuration weights to the
    most recent row of the expanding z-score dataframe.

    Equation
    --------
    s^(k)_t = Σ_i  w_{k,i}  ·  z_{i,t}

    where k ∈ {g, π, pol, fc} denotes the dimension index, i iterates over
    the contributing macro indicators for dimension k, w_{k,i} is the frozen
    weight from STUDY_CONFIG["METHODOLOGY_PARAMS"]["MACRO_SCORING_WEIGHTS"],
    and z_{i,t} is the latest expanding z-score for indicator i.

    Corrections versus prior draft
    --------------------------------
    * Unprescribed ±3 clip_bounds parameter removed.  The manuscript does not
      prescribe clipping dimension scores.  Clipping at ±3 would suppress the
      signal in deep recession or severe inflation scenarios — exactly the
      regimes where maximum discriminative power is most valuable.
    * Hardcoded dimension_mapping dict replaced by the configurable
      dimension_key_mapping parameter (sourced from STUDY_CONFIG at the
      orchestrator layer), eliminating the silent zero-score fallback when
      config keys differ from the hardcoded strings.
    * Missing dimension definition now raises ValueError (fatal), not a
      logger.warning followed by a silent 0.0.
    * pd.Series.at[indicator] used for scalar label-based access — more
      idiomatic and unambiguous than pd.Series.get().
    * NaN z-score raises ValueError rather than being silently replaced with
      0.0.  A NaN at this point indicates a defect in compute_expanding_zscores
      that must be surfaced immediately.
    * Empty df_zscores guard added.

    Parameters
    ----------
    df_zscores : pd.DataFrame
        The expanding z-score dataframe produced by compute_expanding_zscores.
        Must be non-empty.  The most recent row (iloc[-1]) is used.
    scoring_weights : Dict[str, Dict[str, float]]
        Frozen indicator weights per dimension.  Sourced from
        STUDY_CONFIG["METHODOLOGY_PARAMS"]["MACRO_SCORING_WEIGHTS"].
        Structure: {dim_config_key: {indicator_col_name: weight, ...}, ...}.
        Weights within a dimension need NOT sum to 1 — the specification
        prescribes weighted sums, not weighted averages.
    dimension_key_mapping : Dict[str, str]
        Maps dimension config keys (keys of scoring_weights) to the canonical
        output score keys used in the score vector and regime thresholds.
        Must be sourced from
        STUDY_CONFIG["METHODOLOGY_PARAMS"]["MACRO_DIMENSION_KEY_MAP"].
        Example: {"growth": "s_g", "inflation": "s_pi",
                  "monetary_policy": "s_pol", "financial_conditions": "s_fc"}.

    Returns
    -------
    Dict[str, float]
        The four-element dimension score dict with canonical output keys
        (e.g., {"s_g": 0.42, "s_pi": -0.31, "s_pol": 0.78, "s_fc": -0.19}).

    Raises
    ------
    ValueError
        If df_zscores is empty.
        If no weights are defined for any dimension in dimension_key_mapping.
        If a required indicator column is absent from df_zscores.
        If any z-score for a required indicator is NaN at the as-of date.
    """
    # Guard: the z-score dataframe must be non-empty for iloc[-1] to be valid
    if df_zscores.empty:
        raise ValueError(
            "df_zscores is empty.  compute_expanding_zscores must produce "
            "at least one row before compute_dimension_scores is called."
        )

    # Extract the most recent row of z-scores: the macro state at the as-of date
    latest_zscores: pd.Series = df_zscores.iloc[-1]

    # Initialise the output dictionary for the four dimension scores
    dimension_scores: Dict[str, float] = {}

    # Iterate through each dimension in the prescribed mapping order
    for config_key, output_key in dimension_key_mapping.items():

        # Retrieve the weight dictionary for this dimension from the frozen config
        weights: Dict[str, float] = scoring_weights.get(config_key, {})

        # Fatal error if no weights are defined for a dimension — this is a
        # configuration defect that must halt the pipeline, not a warning
        if not weights:
            raise ValueError(
                f"No scoring weights defined for dimension '{config_key}' "
                f"(output key '{output_key}').  "
                f"Add a non-empty weight entry to "
                f"STUDY_CONFIG['METHODOLOGY_PARAMS']['MACRO_SCORING_WEIGHTS']."
            )

        # Pre-flight check: verify every required indicator is present in the z-score frame
        # This surfaces the defect at the dimension level, not inside the inner loop
        missing_indicators: List[str] = [
            ind for ind in weights.keys() if ind not in latest_zscores.index
        ]
        if missing_indicators:
            raise ValueError(
                f"Dimension '{config_key}': the following required indicator(s) are "
                f"absent from the z-score dataframe: {missing_indicators}.  "
                f"Ensure df_macro_raw contains these columns."
            )

        # Initialise the weighted sum accumulator for this dimension
        score: float = 0.0

        # Accumulate the weighted z-score contribution from each indicator
        for indicator, weight in weights.items():

            # Retrieve the scalar z-score using label-based access (.at is faster
            # and unambiguous for single scalar retrieval vs .loc or .get)
            z_val: float = float(latest_zscores.at[indicator])

            # Guard: a NaN z-score at this stage indicates a defect in the expanding
            # window computation that was not caught by the fillna(epsilon) guard.
            # Silently replacing NaN with 0.0 would corrupt the dimension score.
            if np.isnan(z_val):
                raise ValueError(
                    f"NaN z-score detected for indicator '{indicator}' in dimension "
                    f"'{config_key}' at the as-of date.  "
                    f"Investigate compute_expanding_zscores for upstream defects."
                )

            # Accumulate: s^(k)_t += w_{k,i} * z_{i,t}
            score += float(weight) * z_val

        # Store the unclipped weighted-sum score for this dimension.
        # Clipping is deliberately NOT applied: the manuscript does not prescribe it,
        # and clipping would suppress critical information in extreme macro regimes.
        dimension_scores[output_key] = score

    logger.info(
        "compute_dimension_scores: dimension scores = %s",
        {k: round(v, 4) for k, v in dimension_scores.items()},
    )
    # Return the dict of four dimension scores keyed by canonical output labels
    return dimension_scores


# -------------------------------------------------------------------------------------------------------------------------------
# Task 20, Step 2: Compute composite regime score vector and map to 4 regimes
# -------------------------------------------------------------------------------------------------------------------------------

def map_scores_to_regime(
    dimension_scores:  Dict[str, float],
    regime_thresholds: Dict[str, Dict[str, Tuple[Optional[float], Optional[float]]]],
    valid_regimes:     List[str],
) -> Tuple[str, Dict[str, float]]:
    """
    Maps the four-dimensional score vector s_t to a discrete regime label
    via a deterministic, fail-closed exclusive threshold gate, and returns
    the Euclidean distances from s_t to each regime centroid for downstream
    confidence computation.

    Algorithm
    ---------
    1.  Validate that all required dimension score keys are present.
    2.  For each regime r ∈ valid_regimes, evaluate whether the score vector s_t
        jointly satisfies ALL threshold conditions for regime r.
    3.  Count the number of regimes for which all conditions are jointly satisfied.
        * If 0: raise ValueError (score vector escaped all regime boundaries).
        * If >1: raise ValueError (threshold boundaries are non-exclusive —
          configuration defect).
        * If 1: the unique matching regime is selected.
    4.  Compute the centroid of each regime from its threshold parameters.
        For a dimension with both bounds defined: centroid = (lower + upper) / 2.
        For a one-sided bound, the centroid is approximated by offsetting from
        the defined bound by _UNBOUND_REGIME_OFFSET.
    5.  Compute the Euclidean distance from s_t to each regime centroid.
    6.  Return (selected_regime, regime_distances).

    Strict Inequality Convention (per task specification)
    ------------------------------------------------------
    * Lower bound (lower_bound, upper_bound)[0]: EXCLUSIVE.
      Condition: score > lower_bound.  Ensures the lower boundary does not
      simultaneously satisfy two regimes that share the value.
    * Upper bound (lower_bound, upper_bound)[1]: INCLUSIVE.
      Condition: score ≤ upper_bound.  This produces half-open intervals
      (lower, upper] that are mutually exclusive at shared boundary values.

    Corrections versus prior draft
    --------------------------------
    * Soft membership-counting vote replaced by the prescribed deterministic
      exclusive gate (all conditions jointly satisfied).
    * Required fallback ValueError for zero-matching regimes now implemented.
    * Additional ValueError for >1 matching regimes (configuration non-exclusivity).
    * Return type changed from Dict[str, int] (condition counts) to
      Dict[str, float] (Euclidean distances) — geometric distances are the
      prescribed input for the softmax confidence step (Step 3 specification).
    * Regime centroids computed analytically from threshold parameters.

    Parameters
    ----------
    dimension_scores : Dict[str, float]
        The four-element score dict produced by compute_dimension_scores.
        Required keys: the union of all dimension keys referenced in
        regime_thresholds.
    regime_thresholds : Dict[str, Dict[str, Tuple[Optional[float], Optional[float]]]]
        Per-regime, per-dimension (lower, upper) bound pairs.  Sourced from
        STUDY_CONFIG["METHODOLOGY_PARAMS"]["MACRO_REGIME_THRESHOLDS"].
        Structure:
            {
              "Expansion":  {"s_g": (0.5, None), "s_pi": (None, 0.5), ...},
              "Late-cycle": {"s_g": (0.0, 0.5),  "s_pi": (0.5, None), ...},
              ...
            }
        A None lower bound means the dimension is unbounded from below for
        this regime.  A None upper bound means unbounded from above.
    valid_regimes : List[str]
        The canonical ordered list of regime labels:
        ["Expansion", "Late-cycle", "Recession", "Recovery"].
        Evaluation follows this order; the fallback ValueError references it.

    Returns
    -------
    Tuple[str, Dict[str, float]]
        (selected_regime, regime_distances) where:
        * selected_regime: the unique regime label whose threshold conditions
          are all jointly satisfied by the dimension score vector.
        * regime_distances: dict mapping each regime label to its Euclidean
          distance from dimension_scores to the regime's centroid.

    Raises
    ------
    ValueError
        If dimension_scores is missing any key referenced in regime_thresholds.
        If no regime's conditions are jointly satisfied by dimension_scores
        (the score vector escaped all regime boundaries — fail-closed).
        If more than one regime's conditions are jointly satisfied
        (threshold non-exclusivity — configuration defect — fail-closed).
    """
    # Collect every dimension key referenced across all regime threshold dicts
    all_required_keys: Set[str] = set()
    for thresholds in regime_thresholds.values():
        all_required_keys.update(thresholds.keys())

    # Guard: all referenced dimension keys must be present in the score dict
    missing_keys: Set[str] = all_required_keys - set(dimension_scores.keys())
    if missing_keys:
        raise ValueError(
            f"dimension_scores is missing key(s) referenced in regime_thresholds: "
            f"{sorted(missing_keys)}.  "
            f"Ensure compute_dimension_scores produces all required output keys."
        )

    # Initialise the list to accumulate regimes whose conditions are all jointly met
    matching_regimes: List[str] = []

    # Evaluate each regime's threshold conditions against the dimension score vector
    for regime in valid_regimes:

        # Extract the per-dimension threshold conditions for this regime
        thresholds: Dict[str, Tuple[Optional[float], Optional[float]]] = (
            regime_thresholds.get(regime, {})
        )

        # Assume this regime matches until a violated condition falsifies it
        all_conditions_met: bool = True

        # Evaluate each dimension's condition for this regime jointly
        for dim_key, (lower_bound, upper_bound) in thresholds.items():

            # Retrieve the dimension score for this key from the score dict
            score: float = dimension_scores[dim_key]

            # Lower bound: STRICT inequality — score must be strictly ABOVE lower_bound
            # Convention: (lower_bound, upper_bound] half-open interval
            # A None lower_bound means this dimension is unbounded below for this regime
            lower_satisfied: bool = (lower_bound is None) or (score > lower_bound)

            # Upper bound: INCLUSIVE inequality — score must be at or below upper_bound
            # A None upper_bound means this dimension is unbounded above for this regime
            upper_satisfied: bool = (upper_bound is None) or (score <= upper_bound)

            # If either bound is violated, this regime does not match — short-circuit
            if not (lower_satisfied and upper_satisfied):
                all_conditions_met = False
                break

        # Record this regime if all its threshold conditions are jointly satisfied
        if all_conditions_met:
            matching_regimes.append(regime)

    # FALLBACK GATE (per specification Step 2, Omission Prevention):
    # The score vector must match exactly one regime — zero or multiple matches
    # indicate a configuration defect (non-exhaustive or non-exclusive boundaries)
    if len(matching_regimes) == 0:
        raise ValueError(
            f"The dimension score vector {dimension_scores} does not satisfy the "
            f"threshold conditions of ANY regime in valid_regimes={valid_regimes}.  "
            f"The regime_thresholds boundaries are not collectively exhaustive.  "
            f"Revise STUDY_CONFIG['METHODOLOGY_PARAMS']['MACRO_REGIME_THRESHOLDS'] "
            f"to cover the full score space."
        )

    # Multiple-match gate: thresholds must be mutually exclusive
    if len(matching_regimes) > 1:
        raise ValueError(
            f"The dimension score vector {dimension_scores} satisfies the threshold "
            f"conditions of {len(matching_regimes)} regimes simultaneously: "
            f"{matching_regimes}.  "
            f"The regime_thresholds boundaries are not mutually exclusive.  "
            f"Revise STUDY_CONFIG['METHODOLOGY_PARAMS']['MACRO_REGIME_THRESHOLDS'] "
            f"to enforce strict boundary separation."
        )

    # The unique matching regime is the deterministic classification result
    selected_regime: str = matching_regimes[0]

    # ------------------------------------------------------------------
    # STEP 4: Compute regime centroids from threshold parameters
    # ------------------------------------------------------------------
    # For each regime, the centroid in each dimension is the midpoint of
    # the defined bounds.  For half-bounded (one-sided) dimensions the
    # centroid is approximated by offsetting from the defined bound by
    # _UNBOUND_REGIME_OFFSET.  For fully unconstrained dimensions (both
    # bounds None) the centroid is set to 0.0 (the neutral z-score level).

    # Build a dict: {regime_label: {dim_key: centroid_value}}
    regime_centroids: Dict[str, Dict[str, float]] = {}

    for regime in valid_regimes:
        # Retrieve the threshold dict for this regime (empty dict if undefined)
        thresholds = regime_thresholds.get(regime, {})
        centroid_this_regime: Dict[str, float] = {}

        for dim_key in all_required_keys:
            # Retrieve the bounds for this dimension; default to (None, None) if absent
            bounds: Tuple[Optional[float], Optional[float]] = thresholds.get(
                dim_key, (None, None)
            )
            lower_bd: Optional[float] = bounds[0]
            upper_bd: Optional[float] = bounds[1]

            if lower_bd is not None and upper_bd is not None:
                # Both bounds defined: centroid is the arithmetic midpoint of the interval
                centroid_val: float = (lower_bd + upper_bd) / 2.0

            elif lower_bd is not None and upper_bd is None:
                # Only lower bound defined: regime extends to +∞ in this dimension
                # Approximate centroid: lower_bd + offset (2 standard score units above)
                centroid_val = lower_bd + _UNBOUND_REGIME_OFFSET

            elif lower_bd is None and upper_bd is not None:
                # Only upper bound defined: regime extends to −∞ in this dimension
                # Approximate centroid: upper_bd − offset (2 standard score units below)
                centroid_val = upper_bd - _UNBOUND_REGIME_OFFSET

            else:
                # Both bounds are None: dimension is fully unconstrained for this regime
                # Centroid set to neutral z-score level of 0.0
                centroid_val = 0.0

            centroid_this_regime[dim_key] = centroid_val

        regime_centroids[regime] = centroid_this_regime

    # ------------------------------------------------------------------
    # STEP 5: Compute Euclidean distance from s_t to each regime centroid
    # ------------------------------------------------------------------
    # distance_r = sqrt( Σ_d (s_d − centroid_{r,d})^2 )
    # The keys of the distance dict match valid_regimes exactly, providing
    # the geometric proximity signal for the softmax confidence computation.

    regime_distances: Dict[str, float] = {}

    for regime in valid_regimes:
        centroid: Dict[str, float] = regime_centroids[regime]

        # Compute the squared Euclidean distance summed over all dimensions
        squared_distance: float = 0.0
        for dim_key in all_required_keys:
            # Retrieve the score for this dimension from the score dict
            s_d: float = dimension_scores[dim_key]
            # Retrieve the centroid coordinate for this dimension and regime
            c_d: float = centroid[dim_key]
            # Accumulate the squared difference for this dimension
            squared_distance += (s_d - c_d) ** 2.0

        # Take the square root to obtain the Euclidean (L2) distance
        regime_distances[regime] = float(np.sqrt(squared_distance))

    logger.info(
        "map_scores_to_regime: selected_regime='%s' | regime_distances=%s",
        selected_regime,
        {r: round(d, 4) for r, d in regime_distances.items()},
    )
    # Return the selected regime label and the geometric distance vector
    return selected_regime, regime_distances


# -------------------------------------------------------------------------------------------------------------------------------
# Task 20, Step 3: Compute confidence
# -------------------------------------------------------------------------------------------------------------------------------

def compute_softmax_confidence(
    regime_distances: Dict[str, float],
    selected_regime:  str,
    valid_regimes:    List[str],
) -> float:
    """
    Computes a probabilistic confidence score for the selected regime using
    a numerically stable softmax function applied to the NEGATIVE of the
    Euclidean distances from the dimension score vector to each regime
    centroid.

    Equation
    --------
    The softmax is applied to the negated distance vector so that a smaller
    distance (closer regime) yields a higher un-normalised score:

        a_r = −d_r                   (negated distance as activation)

    Numerically stable softmax via log-sum-exp trick (prevents overflow):

        ã_r = a_r − max_i(a_i)       (shift activations by the maximum)
        p_r = exp(ã_r) / Σ_i exp(ã_i)

    The confidence returned is the softmax probability p_selected_regime.

    Corrections versus prior draft
    --------------------------------
    * Parameter renamed from membership_scores: Dict[str, int] to
      regime_distances: Dict[str, float] — distances are geometric continuous
      quantities, as specified by Step 3: "distance of the vector s_t from
      the regime centroids or boundaries."  Integer condition counts are
      not distances and produce inadequate signal sensitivity.
    * Softmax applied to negative distances (closer → higher probability),
      not to raw counts.
    * Prescribed sum-to-1 assertion added:
      assert abs(softmax_probs.sum() - 1.0) < 1e-9.
    * Guard against selected_regime absent from valid_regimes added.
    * Redundant np.clip replaced by a post-hoc assertion — if the extracted
      softmax probability is outside [0, 1], this signals a numerical bug
      that must be surfaced, not silently corrected.

    Parameters
    ----------
    regime_distances : Dict[str, float]
        Euclidean distances from the dimension score vector to each regime
        centroid, produced by map_scores_to_regime.
        Keys must cover all entries in valid_regimes.
    selected_regime : str
        The regime label selected by map_scores_to_regime.
        Must be a member of valid_regimes.
    valid_regimes : List[str]
        The canonical ordered list of regime labels.
        ["Expansion", "Late-cycle", "Recession", "Recovery"].

    Returns
    -------
    float
        The softmax confidence probability for selected_regime, ∈ (0, 1].
        A value of 1.0 indicates the selected regime is the only one
        (distances to all other centroids are +∞).  In practice the value
        will be < 1.0 in proportion to how far the score vector lies from
        alternative regime boundaries.

    Raises
    ------
    ValueError
        If selected_regime is not a member of valid_regimes.
        If regime_distances is missing any key in valid_regimes.
        If the computed confidence is outside [0, 1] after softmax
        (indicates a numerical defect — should not occur under normal conditions).
    AssertionError
        If the softmax probability distribution does not sum to 1.0 within
        1e-9 tolerance (guards against numerical corruption).
    """
    # Guard: selected_regime must be a registered valid regime
    if selected_regime not in valid_regimes:
        raise ValueError(
            f"selected_regime='{selected_regime}' is not in valid_regimes={valid_regimes}.  "
            f"map_scores_to_regime must only return labels from valid_regimes."
        )

    # Guard: all valid regime labels must have a corresponding distance value
    missing_regimes: List[str] = [
        r for r in valid_regimes if r not in regime_distances
    ]
    if missing_regimes:
        raise ValueError(
            f"regime_distances is missing entries for: {missing_regimes}.  "
            f"map_scores_to_regime must return a distance for every regime in "
            f"valid_regimes."
        )

    # Extract distances into a numpy array in the canonical valid_regimes order
    # Canonical ordering ensures the index of selected_regime is consistent
    d_array: np.ndarray = np.array(
        [regime_distances[r] for r in valid_regimes], dtype=np.float64
    )

    # Negate distances to convert "smaller distance → closer" into "higher activation"
    # Softmax amplifies the highest activation, which here corresponds to the
    # closest (most likely) regime
    activation_array: np.ndarray = -d_array

    # Apply the log-sum-exp trick for numerical stability:
    # Subtract the maximum activation before exponentiation to prevent overflow
    # This does not change the softmax output: exp(a - m) / Σ exp(a_i - m) = exp(a) / Σ exp(a_i)
    max_activation: float = float(np.max(activation_array))
    shifted_activations: np.ndarray = activation_array - max_activation

    # Compute the exponentials of the shifted activations
    exp_activations: np.ndarray = np.exp(shifted_activations)

    # Normalise to obtain the softmax probability distribution
    softmax_probs: np.ndarray = exp_activations / np.sum(exp_activations)

    # REQUIRED ASSERTION (per Task 20, Step 3, Omission Prevention):
    # Verify that the softmax probabilities form a valid probability distribution
    # summing to 1.0 within floating-point tolerance
    assert abs(float(softmax_probs.sum()) - 1.0) < 1e-9, (
        f"Softmax probabilities do not sum to 1.0: "
        f"sum={float(softmax_probs.sum()):.12f}."
    )

    # Retrieve the index of the selected regime in the canonical ordering
    selected_idx: int = valid_regimes.index(selected_regime)

    # Extract the softmax probability for the selected regime
    confidence: float = float(softmax_probs[selected_idx])

    # POST-HOC ASSERTION: the extracted confidence must be within [0, 1]
    # This is guaranteed by a correctly computed softmax; if this fails,
    # it indicates a numerical defect that must be surfaced (not clipped)
    if not (0.0 <= confidence <= 1.0):
        raise ValueError(
            f"Computed softmax confidence {confidence:.8f} for '{selected_regime}' "
            f"is outside [0, 1].  This indicates a numerical defect in the "
            f"softmax computation.  Investigate the activation vector: {activation_array}."
        )

    logger.debug(
        "compute_softmax_confidence: regime='%s' | confidence=%.4f | "
        "full_distribution=%s",
        selected_regime,
        confidence,
        {r: round(float(p), 4) for r, p in zip(valid_regimes, softmax_probs)},
    )
    # Return the scalar confidence probability for the selected regime
    return confidence


# -------------------------------------------------------------------------------------------------------------------------------
# Task 20, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_20_macro_scoring_and_classification(
    df_macro_raw: pd.DataFrame,
    config:       Dict[str, Any],
) -> Dict[str, Any]:
    """
    Orchestrates the full macro scoring and regime classification pipeline
    (Task 20).  Chains Steps 1–3 in the prescribed sequential order and
    returns the validated macro_view dictionary for consumption by the Macro
    Agent (Task 17) and downstream CMA and PC pipeline stages.

    Pipeline
    --------
    1. Extract frozen parameters from STUDY_CONFIG.
    2. compute_expanding_zscores  — point-in-time z-scores (fail-closed).
    3. compute_dimension_scores   — four-element score vector s_t.
    4. map_scores_to_regime       — exclusive threshold gate + centroid distances.
    5. compute_softmax_confidence — log-sum-exp softmax on negative distances.
    6. Assemble and validate the macro_view output dict.

    Corrections versus prior draft
    --------------------------------
    * Output schema assertions added: regime ∈ valid_regimes; confidence ∈ [0,1];
      all four dimension score keys present; as_of_date present.
    * Rationale string now correctly references geometric distances rather than
      the defective integer condition counts of the prior draft.
    * dimension_key_mapping extracted from STUDY_CONFIG and threaded through to
      compute_dimension_scores.
    * as_of_date type cast to str explicitly before propagation.

    Parameters
    ----------
    df_macro_raw : pd.DataFrame
        The raw macroeconomic indicator dataframe.  Must have a DatetimeIndex.
        Must contain ONLY observations on or before AS_OF_DATE.
        (compute_expanding_zscores will raise ValueError if future rows exist.)
    config : Dict[str, Any]
        The validated STUDY_CONFIG dictionary.  Required keys:
        - AS_OF_DATE: str (ISO-8601 date)
        - METHODOLOGY_PARAMS.MACRO_SCORING_WEIGHTS: Dict[str, Dict[str, float]]
        - METHODOLOGY_PARAMS.MACRO_REGIME_THRESHOLDS:
              Dict[str, Dict[str, Tuple[Optional[float], Optional[float]]]]
        - METHODOLOGY_PARAMS.MACRO_REGIMES: List[str]
        - METHODOLOGY_PARAMS.MACRO_DIMENSION_KEY_MAP: Dict[str, str]

    Returns
    -------
    Dict[str, Any]
        The validated macro_view dictionary with the following structure:
        {
            "regime":    str,           # one of MACRO_REGIMES
            "scores":    Dict[str, float],   # four dimension scores (s_g, s_pi, etc.)
            "confidence": float,        # softmax confidence ∈ (0, 1]
            "distances": Dict[str, float],   # per-regime centroid distances
            "rationale": str,           # deterministic audit-trail string
            "as_of_date": str,          # frozen AS_OF_DATE from config
        }

    Raises
    ------
    KeyError
        If any required STUDY_CONFIG keys are missing.
    TypeError / ValueError
        Propagated from any sub-function on data or threshold defects.
    AssertionError
        If any output schema post-condition fails.
    """
    logger.info("Task 20: Starting macro scoring and regime classification pipeline.")

    # -----------------------------------------------------------------------
    # Extract all required configuration parameters — fail immediately on any
    # missing key so the error message identifies the exact config path
    # -----------------------------------------------------------------------
    try:
        # Frozen point-in-time date for all expanding window computations
        as_of_date: str = str(config["AS_OF_DATE"])
        # Sub-dict containing all methodology-level parameters
        methodology_params: Dict[str, Any] = config["METHODOLOGY_PARAMS"]
        # Frozen indicator weights per dimension — sourced from Step 1 prescription
        scoring_weights: Dict[str, Dict[str, float]] = methodology_params["MACRO_SCORING_WEIGHTS"]
        # Frozen threshold conditions per regime — sourced from Step 2 prescription
        regime_thresholds: Dict[str, Any] = methodology_params["MACRO_REGIME_THRESHOLDS"]
        # Canonical ordered list of valid regime labels — defines evaluation order
        valid_regimes: List[str] = methodology_params["MACRO_REGIMES"]
        # Mapping from scoring_weights config keys to canonical score vector keys
        # e.g. {"growth": "s_g", "inflation": "s_pi", ...}
        dimension_key_mapping: Dict[str, str] = methodology_params["MACRO_DIMENSION_KEY_MAP"]
    except KeyError as exc:
        raise KeyError(
            f"STUDY_CONFIG is missing a required key for Task 20: {exc}.  "
            f"Ensure METHODOLOGY_PARAMS contains MACRO_SCORING_WEIGHTS, "
            f"MACRO_REGIME_THRESHOLDS, MACRO_REGIMES, and MACRO_DIMENSION_KEY_MAP."
        ) from exc

    # -----------------------------------------------------------------------
    # Step 1: Compute expanding z-scores (point-in-time, fail-closed)
    # -----------------------------------------------------------------------
    logger.debug("Task 20 Step 1: Computing expanding z-scores up to '%s'.", as_of_date)
    df_zscores: pd.DataFrame = compute_expanding_zscores(
        df_macro_raw=df_macro_raw,
        as_of_date=as_of_date,
        # epsilon uses the module default (1e-8); config override not prescribed
    )

    # -----------------------------------------------------------------------
    # Step 1b: Compute the four weighted dimension scores from the z-scores
    # -----------------------------------------------------------------------
    logger.debug("Task 20 Step 1b: Computing dimension scores.")
    dimension_scores: Dict[str, float] = compute_dimension_scores(
        df_zscores=df_zscores,
        scoring_weights=scoring_weights,
        dimension_key_mapping=dimension_key_mapping,
    )

    # -----------------------------------------------------------------------
    # Step 2: Map the dimension scores to a discrete regime (exclusive gate)
    # and compute per-regime centroid distances
    # -----------------------------------------------------------------------
    logger.debug("Task 20 Step 2: Mapping dimension scores to regime.")
    selected_regime: str
    regime_distances: Dict[str, float]
    selected_regime, regime_distances = map_scores_to_regime(
        dimension_scores=dimension_scores,
        regime_thresholds=regime_thresholds,
        valid_regimes=valid_regimes,
    )

    # -----------------------------------------------------------------------
    # Step 3: Compute softmax confidence over geometric centroid distances
    # -----------------------------------------------------------------------
    logger.debug("Task 20 Step 3: Computing softmax confidence.")
    confidence: float = compute_softmax_confidence(
        regime_distances=regime_distances,
        selected_regime=selected_regime,
        valid_regimes=valid_regimes,
    )

    # -----------------------------------------------------------------------
    # Assemble the deterministic rationale string for the audit trail
    # -----------------------------------------------------------------------
    # The rationale references distances (not integer counts) so that it
    # correctly represents the geometric proximity metric used in Step 3
    scores_str: str = ", ".join(
        f"{k}={v:.4f}" for k, v in sorted(dimension_scores.items())
    )
    distances_str: str = ", ".join(
        f"{r}={d:.4f}" for r, d in sorted(regime_distances.items())
    )
    rationale: str = (
        f"Regime classified as '{selected_regime}' with "
        f"{confidence * 100.0:.1f}% softmax confidence (log-sum-exp) "
        f"as of {as_of_date}. "
        f"Dimension Scores: {scores_str}. "
        f"Euclidean distances to regime centroids: {distances_str}. "
        f"Confidence represents the softmax probability of the selected regime "
        f"over the full probability simplex computed from negated centroid distances."
    )

    # -----------------------------------------------------------------------
    # Assemble the macro_view output dict
    # -----------------------------------------------------------------------
    macro_view: Dict[str, Any] = {
        # The discrete regime label — one of the canonical valid_regimes entries
        "regime":     selected_regime,
        # The four-element dimension score vector s_t
        "scores":     dimension_scores,
        # Softmax probability (confidence) for the selected regime ∈ (0, 1]
        "confidence": confidence,
        # Per-regime Euclidean centroid distances — persisted for downstream audit
        "distances":  regime_distances,
        # Deterministic rationale string for fiduciary audit trail
        "rationale":  rationale,
        # Frozen AS_OF_DATE reproduced for provenance
        "as_of_date": as_of_date,
    }

    # -----------------------------------------------------------------------
    # OUTPUT SCHEMA ASSERTIONS — fail-closed post-hoc validation
    # -----------------------------------------------------------------------
    # Assert that the selected regime is a registered valid regime
    assert macro_view["regime"] in valid_regimes, (
        f"Output schema violation: 'regime' value '{macro_view['regime']}' "
        f"is not in valid_regimes={valid_regimes}."
    )

    # Assert that the confidence value is within the valid probability range
    assert 0.0 <= macro_view["confidence"] <= 1.0, (
        f"Output schema violation: 'confidence' value {macro_view['confidence']:.8f} "
        f"is outside [0.0, 1.0]."
    )

    # Assert that all four canonical dimension score keys are present in the output
    canonical_score_keys: Set[str] = set(dimension_key_mapping.values())
    assert canonical_score_keys.issubset(macro_view["scores"].keys()), (
        f"Output schema violation: 'scores' is missing keys "
        f"{canonical_score_keys - set(macro_view['scores'].keys())}."
    )

    # Assert that all valid regime distance keys are present in the distances dict
    assert set(valid_regimes).issubset(macro_view["distances"].keys()), (
        f"Output schema violation: 'distances' is missing regime keys "
        f"{set(valid_regimes) - set(macro_view['distances'].keys())}."
    )

    # Assert the as_of_date field is a non-empty string
    assert isinstance(macro_view["as_of_date"], str) and macro_view["as_of_date"], (
        "Output schema violation: 'as_of_date' must be a non-empty string."
    )

    logger.info(
        "Task 20 completed successfully: regime='%s', confidence=%.4f, as_of='%s'.",
        selected_regime, confidence, as_of_date,
    )
    # Return the fully validated macro_view dictionary
    return macro_view
````

````python
# Task 21: Implement Equity CMA candidate methods 1–7 (script-only computation)

# ==============================================================================
# Task 21: Implement Equity CMA candidate methods 1–7 (script-only computation)
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 21, Step 1: Implement Method 1 (Historical ERP) and Method 2 (Regime-Adjusted ERP)
# -------------------------------------------------------------------------------------------------------------------------------

def execute_cma_method_1_historical_erp(
    monthly_returns: pd.Series,
    rf_series: pd.Series,
    rf_current: float,
    periods_per_year: int,
    min_obs: int = 24,
    epsilon: float = 1e-8,
) -> Dict[str, Any]:
    """
    Computes CMA Method 1: Historical Equity Risk Premium + Current Risk-Free Rate.

    Implements the first candidate expected-return estimator in the seven-method
    CMA pipeline described in Ang, Azimbayev, and Kim (2026), Section 3.2.  This
    function is a pure-Python script callable; it performs no LLM interaction and
    is invoked by the Asset Class (AC) Agent tool-call layer (Task 18).

    Equation (Task 21, Step 1)
    --------------------------
    .. math::

        \\hat{\\mu}_1 = r_f + \\widehat{ERP}_{\\text{hist}}

    where

    .. math::

        \\widehat{ERP}_{\\text{hist}}
            = \\text{periods\\_per\\_year}
              \\times \\overline{(r_{p,t} - r_{f,t})}

    and the overline denotes the arithmetic mean over all aligned observations.

    The annualised information ratio (IR) of the excess-return series is used as
    a monotone proxy for confidence via a logistic transform:

    .. math::

        IR_{\\text{ann}}
            = \\frac{\\widehat{ERP}_{\\text{hist}}}{\\sigma_{\\text{excess,ann}}}

    .. math::

        c_1 = \\sigma\\!\\left(IR_{\\text{ann}}\\right)
            = \\frac{1}{1 + e^{-IR_{\\text{ann}}}}

    A strongly negative IR (i.e., the asset historically underperformed cash)
    correctly drives confidence toward zero; a strongly positive IR drives it
    toward one.

    Pipeline Position
    -----------------
    * Upstream  : historical price/return data loader (Task 18 data-fetch script).
    * Downstream: ``execute_cma_method_7_auto_blend`` (Method 7 aggregator).
    * Output    : one element of the ``method_results`` list written to
                  ``cma_methods.json`` by the orchestrator.

    Parameters
    ----------
    monthly_returns : pd.Series
        Monthly total-return series for the equity asset class expressed as
        **decimal** period returns (e.g., 0.012 for +1.2 %).  Must carry a
        ``pd.DatetimeIndex``.  No NaN values are expected; any NaN rows are
        silently dropped during inner-join alignment with ``rf_series``.
    rf_series : pd.Series
        Monthly risk-free rate series expressed as **decimal** period rates
        (e.g., 0.0035 for 35 bps per month).  Must carry a ``pd.DatetimeIndex``
        aligned to the same frequency as ``monthly_returns``.
    rf_current : float
        The current annualised risk-free rate expressed as a **decimal**
        (e.g., 0.043 for 4.3 %).  Used as the ``r_f`` addend in the equation
        above.  Must satisfy ``|rf_current| <= 1.0``; values outside this range
        are treated as percentage-scale errors and raise ``ValueError``.
    periods_per_year : int
        Number of return observations per calendar year.  Must be 12 for monthly
        data.  Used to annualise both the mean excess return and its standard
        deviation.
    min_obs : int, optional
        Minimum number of aligned (inner-join) observations required before the
        arithmetic mean is considered statistically meaningful.  Default is 24
        (two years of monthly data).  Raises ``ValueError`` if the aligned
        series is shorter.
    epsilon : float, optional
        Small positive floor applied to the annualised excess-return standard
        deviation before division, preventing zero-division in the IR
        calculation.  Default is 1e-8.

    Returns
    -------
    Dict[str, Any]
        A method-result dictionary conforming to the ``cma_methods.json`` output
        contract.  Keys:

        ``method_id`` : int
            Always 1.
        ``estimate`` : float
            Annualised expected return ``r_f + ERP_hist`` expressed as a
            **decimal**.
        ``confidence`` : float
            Logistic-transformed annualised IR; lies in (0, 1).
        ``breakdown`` : Dict[str, Any]
            Component-level audit trail:

            * ``erp_hist``         — annualised historical ERP (decimal).
            * ``rf_current``       — current risk-free rate (decimal).
            * ``ir_annualized``    — annualised information ratio.
            * ``sigma_excess_ann`` — annualised excess-return volatility (decimal).
            * ``n_obs``            — number of aligned observations used.

        ``rationale`` : str
            Human-readable one-line audit string for the CMA Judge skill and
            the board memo.

    Raises
    ------
    TypeError
        If ``monthly_returns`` or ``rf_series`` does not carry a
        ``pd.DatetimeIndex``.
    ValueError
        If ``|rf_current| > 1.0`` (scale guard) or if the number of aligned
        observations is below ``min_obs``.
    """
    # -----------------------------------------------------------------------
    # Input validation: enforce DatetimeIndex on the return series so that
    # the inner-join alignment below is temporally meaningful
    # -----------------------------------------------------------------------
    if not isinstance(monthly_returns.index, pd.DatetimeIndex):
        raise TypeError("monthly_returns must have a pd.DatetimeIndex.")

    # Enforce DatetimeIndex on the risk-free series for the same reason
    if not isinstance(rf_series.index, pd.DatetimeIndex):
        raise TypeError("rf_series must have a pd.DatetimeIndex.")

    # Scale guard: rf_current must be in decimal form; values > 1.0 in absolute
    # value almost certainly indicate a percentage-scale input error
    if abs(rf_current) > 1.0:
        raise ValueError(
            f"rf_current ({rf_current}) appears to be in percentage. "
            "Must be expressed as a decimal (e.g., 0.043 for 4.3 %)."
        )

    # -----------------------------------------------------------------------
    # Temporal alignment: inner join drops any dates present in only one series,
    # ensuring the excess-return computation is element-wise consistent
    # -----------------------------------------------------------------------
    aligned_df: pd.DataFrame = pd.concat(
        [monthly_returns, rf_series], axis=1, join="inner"
    )

    # Assign canonical column names for unambiguous downstream access
    aligned_df.columns = ["r_p", "r_f"]

    # Record the number of aligned observations for the minimum-sample guard
    n_obs: int = len(aligned_df)

    # Enforce minimum observation threshold; fewer observations make the
    # arithmetic mean statistically unreliable as a forward-looking ERP estimate
    if n_obs < min_obs:
        raise ValueError(
            f"Method 1 requires at least {min_obs} aligned observations, "
            f"found {n_obs}."
        )

    # -----------------------------------------------------------------------
    # Core computation: monthly excess-return series r_{p,t} - r_{f,t}
    # -----------------------------------------------------------------------
    excess_returns: pd.Series = aligned_df["r_p"] - aligned_df["r_f"]

    # Arithmetic mean of monthly excess returns — the raw ERP building block
    mean_excess_mo: float = float(excess_returns.mean())

    # Sample standard deviation of monthly excess returns (ddof=1, unbiased)
    std_excess_mo: float = float(excess_returns.std(ddof=1))

    # -----------------------------------------------------------------------
    # Annualisation: scale mean by periods_per_year and std by sqrt thereof
    # Equation: ERP_hist = periods_per_year * mean(r_{p,t} - r_{f,t})
    # -----------------------------------------------------------------------
    erp_hist: float = float(periods_per_year) * mean_excess_mo

    # Annualised excess-return volatility: sigma_ann = sqrt(T) * sigma_monthly
    sigma_excess_ann: float = float(np.sqrt(periods_per_year)) * std_excess_mo

    # -----------------------------------------------------------------------
    # Point estimate: hat{mu}_1 = r_f + ERP_hist
    # -----------------------------------------------------------------------
    estimate: float = rf_current + erp_hist

    # -----------------------------------------------------------------------
    # Confidence scoring via logistic-transformed annualised IR
    # IR_ann = ERP_hist / max(sigma_excess_ann, epsilon)
    # c_1 = 1 / (1 + exp(-IR_ann))
    # Signed IR ensures strongly negative ERP correctly drives confidence to 0
    # -----------------------------------------------------------------------
    ir_annualized: float = erp_hist / max(sigma_excess_ann, epsilon)

    # Logistic (sigmoid) transform maps IR from (-inf, +inf) to (0, 1)
    confidence: float = 1.0 / (1.0 + math.exp(-ir_annualized))

    # -----------------------------------------------------------------------
    # Rationale string: deterministic audit trail for the CMA Judge and board memo
    # -----------------------------------------------------------------------
    rationale: str = (
        f"Method 1 (Historical ERP + Rf): ERP_hist = {erp_hist * 100:.2f}% "
        f"(annualised arithmetic mean over {n_obs} months). "
        f"Current Rf = {rf_current * 100:.2f}%. "
        f"Estimate = {estimate * 100:.2f}%. "
        f"IR_ann = {ir_annualized:.3f} -> Confidence = {confidence:.3f}."
    )

    # Emit a debug-level log entry for pipeline traceability
    logger.debug(
        "Executed Method 1: Estimate=%.4f, Confidence=%.4f", estimate, confidence
    )

    # -----------------------------------------------------------------------
    # Assemble and return the method-result dictionary (cma_methods.json schema)
    # -----------------------------------------------------------------------
    return {
        # Canonical method identifier consumed by Method 7 and the CMA Judge
        "method_id": 1,
        # Annualised expected return expressed as a decimal
        "estimate": float(estimate),
        # Logistic-IR confidence score in (0, 1)
        "confidence": float(confidence),
        # Component-level breakdown for audit and CMA Judge reasoning
        "breakdown": {
            # Annualised historical ERP (decimal)
            "erp_hist": float(erp_hist),
            # Current risk-free rate addend (decimal)
            "rf_current": float(rf_current),
            # Annualised information ratio used to derive confidence
            "ir_annualized": float(ir_annualized),
            # Annualised excess-return volatility (decimal)
            "sigma_excess_ann": float(sigma_excess_ann),
            # Number of aligned monthly observations used
            "n_obs": int(n_obs),
        },
        # Human-readable rationale for the CMA Judge skill and board memo
        "rationale": rationale,
    }


def execute_cma_method_2_regime_adjusted(
    monthly_returns: pd.Series,
    rf_series: pd.Series,
    regime_history: pd.Series,
    current_regime: str,
    regime_confidence: float,
    rf_current: float,
    periods_per_year: int,
    fallback_penalty: float,
    min_regime_obs: int = 12,
) -> Dict[str, Any]:
    """
    Computes CMA Method 2: Regime-Adjusted Equity Risk Premium + Current Risk-Free Rate.

    Implements the second candidate expected-return estimator in the seven-method
    CMA pipeline described in Ang, Azimbayev, and Kim (2026), Section 3.2.  The
    estimator conditions the historical ERP on the current macro regime
    (Expansion, Late-Cycle, Recession, or Recovery) by restricting the mean
    computation to periods whose regime label matches ``current_regime``.

    Equation (Task 21, Step 1)
    --------------------------
    .. math::

        \\hat{\\mu}_2 = r_f + \\widehat{ERP}(\\text{regime})

    where

    .. math::

        \\widehat{ERP}(\\text{regime})
            = \\text{periods\\_per\\_year}
              \\times \\overline{(r_{p,t} - r_{f,t})
                \\;\\big|\\; \\text{regime}_t = \\text{current\\_regime}}

    Fallback Behaviour
    ------------------
    If the number of regime-matching observations is below ``min_regime_obs``,
    the function falls back to the unconditional mean (identical to Method 1's
    ERP) and applies ``fallback_penalty`` as a multiplicative confidence
    discount, reflecting the reduced informativeness of the estimate.

    Confidence Scoring
    ------------------
    .. math::

        c_2 = \\text{clip}\\!\\left(
                  \\hat{c}_{\\text{regime}} \\times q,\\; 0,\\; 1
              \\right)

    where :math:`\\hat{c}_{\\text{regime}}` is the macro agent's regime
    classification confidence (clipped to [0, 1]) and :math:`q` is a sample
    quality ramp:

    * If sufficient regime observations exist:
      :math:`q = \\min\\!\\left(1,\\; n_{\\text{regime}} /
      (2 \\times \\text{min\\_regime\\_obs})\\right)`.
    * If the fallback is triggered: :math:`q = \\text{fallback\\_penalty}`.

    Pipeline Position
    -----------------
    * Upstream  : macro regime history series (Task 20 output) and historical
                  return data loader (Task 18 data-fetch script).
    * Downstream: ``execute_cma_method_7_auto_blend`` (Method 7 aggregator).
    * Output    : one element of the ``method_results`` list written to
                  ``cma_methods.json`` by the orchestrator.

    Parameters
    ----------
    monthly_returns : pd.Series
        Monthly total-return series for the equity asset class expressed as
        **decimal** period returns.  Must carry a ``pd.DatetimeIndex``.
    rf_series : pd.Series
        Monthly risk-free rate series expressed as **decimal** period rates.
        Must carry a ``pd.DatetimeIndex`` aligned to the same frequency as
        ``monthly_returns``.
    regime_history : pd.Series
        Monthly regime-label series (str dtype) aligned to the same
        ``pd.DatetimeIndex`` as ``monthly_returns``.  Labels must be drawn from
        the canonical set: ``{"Expansion", "Late-Cycle", "Recession",
        "Recovery"}``.
    current_regime : str
        The regime label produced by the macro agent (Task 20) for the current
        period.  Used as the Boolean mask key against ``regime_history``.
    regime_confidence : float
        The macro agent's softmax confidence for ``current_regime``, expressed
        in [0, 1].  Values outside this range are clipped with a warning.
    rf_current : float
        The current annualised risk-free rate expressed as a **decimal**.
    periods_per_year : int
        Number of return observations per calendar year (12 for monthly data).
    fallback_penalty : float
        Multiplicative confidence discount applied when the fallback to the
        unconditional mean is triggered.  Must lie in (0, 1].  Sourced from
        ``STUDY_CONFIG["METHODOLOGY_PARAMS"]["CMA_METHOD_2_FALLBACK_PENALTY"]``.
    min_regime_obs : int, optional
        Minimum number of regime-matching observations required before the
        conditional mean is used.  Default is 12 (one year of monthly data).

    Returns
    -------
    Dict[str, Any]
        A method-result dictionary conforming to the ``cma_methods.json`` output
        contract.  Keys:

        ``method_id`` : int
            Always 2.
        ``estimate`` : float
            Annualised expected return ``r_f + ERP_regime`` expressed as a
            **decimal**.
        ``confidence`` : float
            Regime-confidence-weighted sample-quality score; lies in [0, 1].
        ``breakdown`` : Dict[str, Any]
            Component-level audit trail:

            * ``erp_regime``        — annualised regime-conditional ERP (decimal).
            * ``rf_current``        — current risk-free rate (decimal).
            * ``n_regime_obs``      — number of regime-matching observations.
            * ``regime_confidence`` — clipped macro-agent confidence.
            * ``fallback_used``     — True if the unconditional fallback was used.

        ``rationale`` : str
            Human-readable one-line audit string for the CMA Judge skill.

    Raises
    ------
    TypeError
        If any of ``monthly_returns``, ``rf_series``, or ``regime_history``
        does not carry a ``pd.DatetimeIndex``.
    ValueError
        If ``rf_current`` is outside the decimal scale guard (|rf_current| > 1).
    """
    # -----------------------------------------------------------------------
    # Temporal alignment: inner join across all three series so that every
    # row used in the conditional mean has a contemporaneous return, rf, and
    # regime label — preventing any look-ahead or missing-data contamination
    # -----------------------------------------------------------------------
    aligned_df: pd.DataFrame = pd.concat(
        [monthly_returns, rf_series, regime_history], axis=1, join="inner"
    )

    # Assign canonical column names for unambiguous downstream access
    aligned_df.columns = ["r_p", "r_f", "regime"]

    # -----------------------------------------------------------------------
    # Boolean mask: retain only rows whose regime label matches current_regime
    # Implements the conditioning: E[r_p - r_f | regime_t = current_regime]
    # -----------------------------------------------------------------------
    regime_df: pd.DataFrame = aligned_df[aligned_df["regime"] == current_regime]

    # Record the number of regime-matching observations for the minimum-sample guard
    n_regime_obs: int = len(regime_df)

    # Unconditionally log the observation count to satisfy the Task 21 Step 1
    # omission-prevention requirement: "Explicitly log the number of observations
    # used for the regime-adjusted calculation."
    logger.info(
        "Method 2 regime '%s' observation count: %d", current_regime, n_regime_obs
    )

    # Initialise fallback flag; will be set True if the threshold is not met
    fallback_used: bool = False

    if n_regime_obs >= min_regime_obs:
        # Sufficient regime-specific data: compute the conditional excess-return mean
        excess_returns: pd.Series = regime_df["r_p"] - regime_df["r_f"]

        # Arithmetic mean of monthly excess returns within the current regime
        mean_excess_mo: float = float(excess_returns.mean())

        # Sample quality ramp: linearly scales from 0 at min_regime_obs to 1 at
        # 2 * min_regime_obs, capped at 1.0 for larger samples
        sample_factor: float = min(
            1.0, float(n_regime_obs) / float(min_regime_obs * 2)
        )

    else:
        # Insufficient regime-specific data: fall back to the unconditional mean
        # and apply the configured confidence penalty to signal reduced reliability
        fallback_used = True

        logger.warning(
            "Method 2: Insufficient observations (%d < %d) for regime '%s'. "
            "Falling back to unconditional mean with penalty %.3f.",
            n_regime_obs,
            min_regime_obs,
            current_regime,
            fallback_penalty,
        )

        # Unconditional excess-return series across all aligned observations
        excess_returns_full: pd.Series = aligned_df["r_p"] - aligned_df["r_f"]

        # Arithmetic mean of the full (unconditional) excess-return series
        mean_excess_mo = float(excess_returns_full.mean())

        # Apply the configured fallback penalty as the sample quality factor
        sample_factor = float(fallback_penalty)

    # -----------------------------------------------------------------------
    # Annualisation: ERP_regime = periods_per_year * mean(r_{p,t} - r_{f,t} | regime)
    # -----------------------------------------------------------------------
    erp_regime: float = float(periods_per_year) * mean_excess_mo

    # Point estimate: hat{mu}_2 = r_f + ERP_regime
    estimate: float = rf_current + erp_regime

    # -----------------------------------------------------------------------
    # Confidence scoring: c_2 = clip(regime_confidence * sample_factor, 0, 1)
    # Clip regime_confidence to [0, 1] before multiplication to guard against
    # out-of-range macro-agent outputs
    # -----------------------------------------------------------------------
    safe_regime_conf: float = float(np.clip(regime_confidence, 0.0, 1.0))

    # Final confidence: product of macro-agent regime confidence and sample quality
    confidence: float = float(np.clip(safe_regime_conf * sample_factor, 0.0, 1.0))

    # -----------------------------------------------------------------------
    # Rationale string: deterministic audit trail for the CMA Judge and board memo
    # -----------------------------------------------------------------------
    fallback_str: str = " (Fallback to full history used)" if fallback_used else ""

    rationale: str = (
        f"Method 2 (Regime-Adjusted ERP + Rf): "
        f"Current regime = '{current_regime}' (conf={safe_regime_conf:.3f}). "
        f"Found {n_regime_obs} matching periods{fallback_str}. "
        f"ERP_regime = {erp_regime * 100:.2f}%. "
        f"Estimate = {estimate * 100:.2f}%. "
        f"Confidence = {confidence:.3f}."
    )

    # Emit a debug-level log entry for pipeline traceability
    logger.debug(
        "Executed Method 2: Estimate=%.4f, Confidence=%.4f", estimate, confidence
    )

    # -----------------------------------------------------------------------
    # Assemble and return the method-result dictionary (cma_methods.json schema)
    # -----------------------------------------------------------------------
    return {
        # Canonical method identifier consumed by Method 7 and the CMA Judge
        "method_id": 2,
        # Annualised expected return expressed as a decimal
        "estimate": float(estimate),
        # Regime-confidence-weighted sample-quality score in [0, 1]
        "confidence": float(confidence),
        # Component-level breakdown for audit and CMA Judge reasoning
        "breakdown": {
            # Annualised regime-conditional ERP (decimal)
            "erp_regime": float(erp_regime),
            # Current risk-free rate addend (decimal)
            "rf_current": float(rf_current),
            # Number of regime-matching monthly observations used
            "n_regime_obs": int(n_regime_obs),
            # Clipped macro-agent regime classification confidence
            "regime_confidence": float(safe_regime_conf),
            # Flag indicating whether the unconditional fallback was triggered
            "fallback_used": bool(fallback_used),
        },
        # Human-readable rationale for the CMA Judge skill and board memo
        "rationale": rationale,
    }


# -------------------------------------------------------------------------------------------------------------------------------
# Task 21, Step 2: Implement Method 3 (Black-Litterman Equilibrium Prior)
# -------------------------------------------------------------------------------------------------------------------------------

def execute_cma_method_3_bl_equilibrium(
    asset_index: int,
    sigma: np.ndarray,
    w_mkt: np.ndarray,
    bl_delta: float = 2.5,
    epsilon: float = 1e-8,
) -> Dict[str, Any]:
    """
    Computes CMA Method 3: Black-Litterman Equilibrium Implied Return.

    Implements the third candidate expected-return estimator in the seven-method
    CMA pipeline described in Ang, Azimbayev, and Kim (2026), Section 3.2.  The
    estimator derives the market-implied (equilibrium) expected return vector
    from the canonical Black-Litterman (1992) prior, then extracts the scalar
    estimate for the target asset class.

    Equation (Task 21, Step 2)
    --------------------------
    .. math::

        \\pi = \\delta \\, \\Sigma \\, w_{\\text{mkt}}

    where:

    * :math:`\\delta` is the market risk-aversion coefficient (frozen at 2.5
      per the Task 21 specification).
    * :math:`\\Sigma` is the :math:`18 \\times 18` annualised asset-class
      covariance matrix produced by the Covariance Agent (Task 21, Step 2).
    * :math:`w_{\\text{mkt}}` is the :math:`18 \\times 1` vector of
      market-capitalisation weights summing to 1.

    The scalar estimate for asset ``asset_index`` is
    :math:`\\hat{\\mu}_3 = \\pi[\\text{asset\\_index}]`.

    Confidence Scoring
    ------------------
    Confidence is proportional to the asset's market-cap weight relative to
    the maximum weight in the universe, reflecting the intuition that
    equilibrium priors are most reliable for large, liquid, well-priced assets:

    .. math::

        c_3 = \\frac{w_{\\text{mkt},i}}{\\max_j w_{\\text{mkt},j} + \\epsilon}

    Pipeline Position
    -----------------
    * Upstream  : Covariance Agent (Task 21, Step 2) for :math:`\\Sigma`;
                  market-cap weight vector from the data-fetch script.
    * Downstream: ``execute_cma_method_7_auto_blend`` (Method 7 aggregator).
    * Output    : one element of the ``method_results`` list written to
                  ``cma_methods.json`` by the orchestrator.

    Parameters
    ----------
    asset_index : int
        Zero-based index of the target asset class within the 18-asset universe
        defined in ``STUDY_CONFIG["IPS_GOVERNANCE"]["ASSET_UNIVERSE"]``.
        Must satisfy ``0 <= asset_index < 18``.
    sigma : np.ndarray
        Annualised asset-class covariance matrix of shape ``(18, 18)``.  Must
        be positive semi-definite (PSD); the function checks the minimum
        eigenvalue and raises ``ValueError`` if it is below -1e-5.
    w_mkt : np.ndarray
        Market-capitalisation weight vector of shape ``(18,)``.  Must sum to
        1.0 within a tolerance of 1e-4 and must contain no negative entries.
    bl_delta : float, optional
        Market risk-aversion coefficient :math:`\\delta`.  Frozen at 2.5 per
        the Task 21 specification.  Default is 2.5.
    epsilon : float, optional
        Small positive floor applied to the denominator of the confidence
        ratio, preventing zero-division when all market weights are zero.
        Default is 1e-8.

    Returns
    -------
    Dict[str, Any]
        A method-result dictionary conforming to the ``cma_methods.json`` output
        contract.  Keys:

        ``method_id`` : int
            Always 3.
        ``estimate`` : float
            Annualised equilibrium expected return
            :math:`\\pi[\\text{asset\\_index}]` expressed as a **decimal**.
        ``confidence`` : float
            Market-weight-proportional confidence score; lies in [0, 1].
        ``breakdown`` : Dict[str, Any]
            Component-level audit trail:

            * ``delta``               — risk-aversion coefficient used.
            * ``w_mkt_i``             — market-cap weight of the target asset.
            * ``w_mkt_max``           — maximum market-cap weight in the universe.
            * ``sigma_min_eigenvalue``— minimum eigenvalue of :math:`\\Sigma`.

        ``rationale`` : str
            Human-readable one-line audit string for the CMA Judge skill.

    Raises
    ------
    ValueError
        If ``sigma`` is not of shape ``(18, 18)``, if ``w_mkt`` is not of
        shape ``(18,)``, if ``w_mkt`` does not sum to 1.0 within tolerance,
        if ``w_mkt`` contains negative entries, or if ``sigma`` is not PSD.
    IndexError
        If ``asset_index`` is outside the range ``[0, 17]``.
    """
    # -----------------------------------------------------------------------
    # Dimensional guard: sigma must be exactly 18 x 18 to match the IPS universe
    # Task 21 Step 2 omission-prevention: "Assert that Sigma and w_mkt are
    # perfectly aligned dimensionally (18 x 18 and 18 x 1)."
    # -----------------------------------------------------------------------
    if sigma.shape != (18, 18):
        raise ValueError(
            f"Method 3 requires an 18x18 covariance matrix, got {sigma.shape}."
        )

    # Dimensional guard: w_mkt must be a flat 18-element vector
    if w_mkt.shape != (18,):
        raise ValueError(
            f"Method 3 requires an 18-element market weight vector, got {w_mkt.shape}."
        )

    # Bounds guard: asset_index must be a valid zero-based index into the universe
    if not (0 <= asset_index < 18):
        raise IndexError(
            f"asset_index {asset_index} is out of bounds for the 18-asset universe."
        )

    # -----------------------------------------------------------------------
    # Market-weight integrity checks: strict sum-to-1 and non-negativity
    # Silent normalisation is deliberately avoided; a violated sum indicates
    # a data-pipeline error that must be surfaced immediately
    # -----------------------------------------------------------------------
    w_sum: float = float(w_mkt.sum())

    # Enforce that market weights sum to 1.0 within a 1e-4 tolerance
    if abs(w_sum - 1.0) > 1e-4:
        raise ValueError(
            f"Market weight vector must sum to 1.0, but sums to {w_sum:.6f}."
        )

    # Enforce non-negativity: negative market-cap weights are economically incoherent
    if np.any(w_mkt < 0.0):
        raise ValueError(
            "Market weight vector contains negative weights, which is economically incoherent."
        )

    # -----------------------------------------------------------------------
    # PSD check: compute the minimum eigenvalue of sigma via symmetric solver
    # A minimum eigenvalue below -1e-5 indicates numerical non-PSD, which
    # would make the BL equilibrium prior ill-defined
    # -----------------------------------------------------------------------
    min_eigenvalue: float = float(np.linalg.eigvalsh(sigma).min())

    if min_eigenvalue < -1e-5:
        raise ValueError(
            f"Covariance matrix is not PSD. "
            f"Minimum eigenvalue: {min_eigenvalue:.6e}. "
            "Apply a Ledoit-Wolf or nearest-PSD correction before calling Method 3."
        )

    # -----------------------------------------------------------------------
    # Core computation: pi = delta * Sigma * w_mkt
    # Black-Litterman (1992) equilibrium implied return vector
    # Equation: pi = delta * Sigma @ w_mkt
    # -----------------------------------------------------------------------
    pi_vector: np.ndarray = bl_delta * (sigma @ w_mkt)

    # Extract the scalar estimate for the target asset class
    estimate: float = float(pi_vector[asset_index])

    # -----------------------------------------------------------------------
    # Confidence scoring: c_3 = w_mkt_i / max(w_mkt_max, epsilon)
    # Proportional to the asset's market-cap weight relative to the largest
    # weight in the universe; large, liquid assets receive higher confidence
    # -----------------------------------------------------------------------
    w_mkt_i: float = float(w_mkt[asset_index])

    # Maximum market-cap weight across the full 18-asset universe
    w_mkt_max: float = float(w_mkt.max())

    # Confidence ratio: normalised by the maximum weight, floored by epsilon
    confidence: float = float(w_mkt_i / max(w_mkt_max, epsilon))

    # -----------------------------------------------------------------------
    # Rationale string: deterministic audit trail for the CMA Judge and board memo
    # -----------------------------------------------------------------------
    rationale: str = (
        f"Method 3 (BL Equilibrium Prior): pi = delta * Sigma * w_mkt. "
        f"delta = {bl_delta:.2f}. "
        f"Asset market-cap weight = {w_mkt_i * 100:.2f}%. "
        f"Estimate = {estimate * 100:.2f}%. "
        f"Confidence (weight-proportional) = {confidence:.3f}."
    )

    # Emit a debug-level log entry for pipeline traceability
    logger.debug(
        "Executed Method 3: Estimate=%.4f, Confidence=%.4f", estimate, confidence
    )

    # -----------------------------------------------------------------------
    # Assemble and return the method-result dictionary (cma_methods.json schema)
    # -----------------------------------------------------------------------
    return {
        # Canonical method identifier consumed by Method 7 and the CMA Judge
        "method_id": 3,
        # Annualised equilibrium expected return expressed as a decimal
        "estimate": float(estimate),
        # Market-weight-proportional confidence score in [0, 1]
        "confidence": float(confidence),
        # Component-level breakdown for audit and CMA Judge reasoning
        "breakdown": {
            # Risk-aversion coefficient delta used in the BL prior
            "delta": float(bl_delta),
            # Market-cap weight of the target asset class
            "w_mkt_i": float(w_mkt_i),
            # Maximum market-cap weight in the 18-asset universe
            "w_mkt_max": float(w_mkt_max),
            # Minimum eigenvalue of sigma — PSD diagnostic
            "sigma_min_eigenvalue": float(min_eigenvalue),
        },
        # Human-readable rationale for the CMA Judge skill and board memo
        "rationale": rationale,
    }


# -------------------------------------------------------------------------------------------------------------------------------
# Task 21, Step 3: Implement Method 4 (Inverse Gordon) and Method 5 (Implied ERP)
# -------------------------------------------------------------------------------------------------------------------------------

def execute_cma_method_4_inverse_gordon(
    dividend_yield: float,
    buyback_yield: float,
    earnings_growth: float,
    valuation_change: float,
    conf_explicit: float,
    conf_proxy: float,
    growth_is_proxy: bool = False,
) -> Dict[str, Any]:
    """
    Computes CMA Method 4: Inverse Gordon / Grinold-Kroner Building-Block Model.

    Implements the fourth candidate expected-return estimator in the seven-method
    CMA pipeline described in Ang, Azimbayev, and Kim (2026), Section 3.2.  The
    estimator decomposes the forward equity return into three fundamental
    building blocks following Gordon (1959) and Grinold and Kroner (2002):
    income yield, earnings growth, and valuation change.

    Equation (Task 21, Step 2)
    --------------------------
    .. math::

        \\hat{\\mu}_4
            = y_{\\text{div}} + y_{\\text{buyback}} + g + \\Delta v

    where:

    * :math:`y_{\\text{div}}`     — dividend yield (decimal).
    * :math:`y_{\\text{buyback}}` — net share buyback yield (decimal).
    * :math:`g`                   — expected earnings growth rate (decimal);
      sourced from analyst forecasts or, if unavailable, from a nominal GDP
      growth proxy.
    * :math:`\\Delta v`           — expected valuation change (decimal);
      positive implies multiple expansion, negative implies compression.

    Fallback Behaviour
    ------------------
    If ``growth_is_proxy`` is ``True``, the earnings growth input was derived
    from a nominal GDP proxy rather than a direct analyst forecast.  In this
    case the lower confidence score ``conf_proxy`` is applied, reflecting the
    reduced precision of the proxy.  This implements the Task 21 Step 2
    resolution: "implementing a fallback to a nominal GDP growth proxy (e.g.,
    4.0 %), applying a confidence penalty when the proxy is used."

    Pipeline Position
    -----------------
    * Upstream  : fundamental data loader (``df_fundamentals_raw``) in the
                  orchestrator; ``earnings_growth_forecast`` column; nominal
                  GDP proxy from ``STUDY_CONFIG``.
    * Downstream: ``execute_cma_method_7_auto_blend`` (Method 7 aggregator).
    * Output    : one element of the ``method_results`` list written to
                  ``cma_methods.json`` by the orchestrator.

    Parameters
    ----------
    dividend_yield : float
        Trailing or forward dividend yield expressed as a **decimal**
        (e.g., 0.013 for 1.3 %).  Must be non-negative and satisfy
        ``|dividend_yield| <= 1.0``.
    buyback_yield : float
        Net share buyback yield expressed as a **decimal**
        (e.g., 0.018 for 1.8 %).  Must be non-negative and satisfy
        ``|buyback_yield| <= 1.0``.
    earnings_growth : float
        Expected earnings growth rate expressed as a **decimal**
        (e.g., 0.05 for 5.0 %).  May be negative (earnings contraction).
        Must satisfy ``|earnings_growth| <= 1.0``.  If sourced from a nominal
        GDP proxy, set ``growth_is_proxy=True``.
    valuation_change : float
        Expected change in the price-to-earnings multiple expressed as a
        **decimal** (e.g., -0.01 for -1.0 % valuation compression).  Must
        satisfy ``|valuation_change| <= 1.0``.
    conf_explicit : float
        Confidence score applied when ``earnings_growth`` is sourced from an
        explicit analyst forecast.  Sourced from
        ``STUDY_CONFIG["METHODOLOGY_PARAMS"]["CMA_METHOD_4_CONF_EXPLICIT"]``.
        Must lie in [0, 1].
    conf_proxy : float
        Confidence score applied when ``earnings_growth`` is sourced from the
        nominal GDP proxy.  Sourced from
        ``STUDY_CONFIG["METHODOLOGY_PARAMS"]["CMA_METHOD_4_CONF_PROXY"]``.
        Must lie in [0, 1] and is expected to be strictly less than
        ``conf_explicit``.
    growth_is_proxy : bool, optional
        Flag indicating whether ``earnings_growth`` was derived from the
        nominal GDP proxy rather than a direct analyst forecast.  Default is
        ``False``.

    Returns
    -------
    Dict[str, Any]
        A method-result dictionary conforming to the ``cma_methods.json`` output
        contract.  Keys:

        ``method_id`` : int
            Always 4.
        ``estimate`` : float
            Annualised expected return
            :math:`y_{\\text{div}} + y_{\\text{buyback}} + g + \\Delta v`
            expressed as a **decimal**.
        ``confidence`` : float
            Either ``conf_explicit`` or ``conf_proxy`` depending on
            ``growth_is_proxy``; lies in [0, 1].
        ``breakdown`` : Dict[str, Any]
            Component-level audit trail:

            * ``dividend_yield``    — dividend yield component (decimal).
            * ``buyback_yield``     — buyback yield component (decimal).
            * ``earnings_growth``   — growth component (decimal).
            * ``valuation_change``  — valuation change component (decimal).
            * ``growth_is_proxy``   — True if the GDP proxy was used.

        ``rationale`` : str
            Human-readable one-line audit string for the CMA Judge skill.

    Raises
    ------
    ValueError
        If any of the four yield/growth/valuation inputs has an absolute value
        exceeding 1.0 (decimal scale guard), or if ``dividend_yield`` or
        ``buyback_yield`` is negative.
    """
    # -----------------------------------------------------------------------
    # Decimal scale guard: all four building-block inputs must be in decimal
    # form; values with absolute magnitude > 1.0 almost certainly indicate a
    # percentage-scale input error that would produce a nonsensical estimate
    # -----------------------------------------------------------------------
    for name, val in [
        ("dividend_yield", dividend_yield),
        ("buyback_yield", buyback_yield),
        ("earnings_growth", earnings_growth),
        ("valuation_change", valuation_change),
    ]:
        if abs(val) > 1.0:
            raise ValueError(
                f"Fatal Scale Error: {name} ({val}) exceeds the 1.0 absolute "
                "limit. All building-block inputs must be expressed as decimals "
                "(e.g., 0.013 for 1.3 %)."
            )

    # Non-negativity guard: dividend and buyback yields cannot be negative
    # (a negative dividend yield is economically incoherent for a long-only asset)
    if dividend_yield < 0.0 or buyback_yield < 0.0:
        raise ValueError(
            "dividend_yield and buyback_yield must be non-negative. "
            f"Received dividend_yield={dividend_yield}, buyback_yield={buyback_yield}."
        )

    # -----------------------------------------------------------------------
    # Core computation: hat{mu}_4 = y_div + y_buyback + g + Delta_v
    # Equation: Grinold-Kroner (2002) building-block decomposition
    # -----------------------------------------------------------------------
    estimate: float = (
        dividend_yield + buyback_yield + earnings_growth + valuation_change
    )

    # -----------------------------------------------------------------------
    # Confidence selection: apply the proxy penalty if earnings_growth was
    # derived from the nominal GDP proxy rather than an analyst forecast
    # -----------------------------------------------------------------------
    confidence: float = conf_proxy if growth_is_proxy else conf_explicit

    # -----------------------------------------------------------------------
    # Rationale string: deterministic audit trail for the CMA Judge and board memo
    # Annotate the growth component with "(Nominal GDP Proxy)" when applicable
    # -----------------------------------------------------------------------
    proxy_str: str = " (Nominal GDP Proxy)" if growth_is_proxy else ""

    rationale: str = (
        f"Method 4 (Inverse Gordon / Grinold-Kroner): "
        f"mu = y_div + y_buyback + g + Delta_v. "
        f"y_div={dividend_yield * 100:.2f}%, "
        f"y_buyback={buyback_yield * 100:.2f}%, "
        f"g={earnings_growth * 100:.2f}%{proxy_str}, "
        f"Delta_v={valuation_change * 100:.2f}%. "
        f"Estimate = {estimate * 100:.2f}%. "
        f"Confidence = {confidence:.3f}."
    )

    # Emit a debug-level log entry for pipeline traceability
    logger.debug(
        "Executed Method 4: Estimate=%.4f, Confidence=%.4f", estimate, confidence
    )

    # -----------------------------------------------------------------------
    # Assemble and return the method-result dictionary (cma_methods.json schema)
    # -----------------------------------------------------------------------
    return {
        # Canonical method identifier consumed by Method 7 and the CMA Judge
        "method_id": 4,
        # Annualised expected return expressed as a decimal
        "estimate": float(estimate),
        # Confidence score: conf_explicit or conf_proxy depending on data quality
        "confidence": float(confidence),
        # Component-level breakdown for audit and CMA Judge reasoning
        "breakdown": {
            # Dividend yield building block (decimal)
            "dividend_yield": float(dividend_yield),
            # Net buyback yield building block (decimal)
            "buyback_yield": float(buyback_yield),
            # Earnings growth building block (decimal); may be a GDP proxy
            "earnings_growth": float(earnings_growth),
            # Valuation change building block (decimal)
            "valuation_change": float(valuation_change),
            # Flag indicating whether the GDP proxy was used for earnings growth
            "growth_is_proxy": bool(growth_is_proxy),
        },
        # Human-readable rationale for the CMA Judge skill and board memo
        "rationale": rationale,
    }


def execute_cma_method_5_implied_erp(
    earnings_yield: float,
    rf_current: float,
    cape_current: float,
    cape_history: Optional[pd.Series] = None,
) -> Dict[str, Any]:
    """
    Computes CMA Method 5: Implied ERP Proxy via Campbell-Shiller CAPE Earnings Yield.

    Implements the fifth candidate expected-return estimator in the seven-method
    CMA pipeline described in Ang, Azimbayev, and Kim (2026), Section 3.2.  The
    estimator uses the cyclically adjusted earnings yield (the reciprocal of the
    Campbell-Shiller CAPE ratio) as a proxy for the forward equity return, and
    derives the implied equity risk premium by subtracting the current risk-free
    rate.

    Equations (Task 21, Step 2)
    ---------------------------
    Implied ERP proxy:

    .. math::

        \\widehat{ERP}_{\\text{impl}}
            \\approx \\text{EarningsYield} - r_f
            = \\frac{E_{10}}{P} - r_f

    Point estimate (the earnings yield itself serves as the total return proxy):

    .. math::

        \\hat{\\mu}_5 = \\text{EarningsYield} = \\frac{1}{\\text{CAPE}}

    Confidence Scoring
    ------------------
    Confidence is inversely proportional to the CAPE percentile within its
    historical distribution.  A high CAPE (expensive market) implies that the
    earnings yield is a less reliable forward-return predictor (mean-reversion
    risk is elevated), so confidence is penalised:

    .. math::

        c_5 = 1 - \\text{Percentile}(\\text{CAPE}_{\\text{current}},
                                       \\text{CAPE}_{\\text{history}})

    If ``cape_history`` is unavailable, the percentile defaults to 0.5
    (neutral confidence of 0.5).

    Pipeline Position
    -----------------
    * Upstream  : fundamental data loader (``df_fundamentals_raw``) in the
                  orchestrator; ``earnings_yield`` and ``cape_ratio`` columns.
    * Downstream: ``execute_cma_method_7_auto_blend`` (Method 7 aggregator).
    * Output    : one element of the ``method_results`` list written to
                  ``cma_methods.json`` by the orchestrator.

    Parameters
    ----------
    earnings_yield : float
        Cyclically adjusted earnings yield expressed as a **decimal**
        (e.g., 0.032 for 3.2 %).  Typically computed as ``1 / CAPE``.
        Must satisfy ``|earnings_yield| <= 1.0``.
    rf_current : float
        The current annualised risk-free rate expressed as a **decimal**
        (e.g., 0.037 for 3.7 %).  Must satisfy ``|rf_current| <= 1.0``.
    cape_current : float
        The current Cyclically Adjusted Price-to-Earnings ratio (CAPE / Shiller
        PE10).  Must be strictly positive (e.g., 25.0 for a CAPE of 25x).
    cape_history : pd.Series, optional
        Historical time series of CAPE values used to compute the current
        CAPE's percentile rank within its own distribution.  If ``None`` or
        empty, the percentile defaults to 0.5 and a warning is logged.

    Returns
    -------
    Dict[str, Any]
        A method-result dictionary conforming to the ``cma_methods.json`` output
        contract.  Keys:

        ``method_id`` : int
            Always 5.
        ``estimate`` : float
            Annualised expected return (earnings yield) expressed as a
            **decimal**.
        ``confidence`` : float
            CAPE-percentile-inverted confidence score; lies in [0, 1].
        ``breakdown`` : Dict[str, Any]
            Component-level audit trail:

            * ``earnings_yield``  — CAPE-based earnings yield (decimal).
            * ``rf_current``      — current risk-free rate (decimal).
            * ``erp_implied``     — implied ERP = earnings_yield - rf (decimal).
            * ``cape_current``    — current CAPE ratio (multiple).
            * ``cape_percentile`` — CAPE percentile within historical distribution.

        ``rationale`` : str
            Human-readable one-line audit string for the CMA Judge skill.

    Raises
    ------
    ValueError
        If ``|earnings_yield| > 1.0`` or ``|rf_current| > 1.0`` (scale guards),
        or if ``cape_current`` is non-positive or NaN.
    """
    # -----------------------------------------------------------------------
    # Decimal scale guard: earnings_yield and rf_current must be in decimal form
    # -----------------------------------------------------------------------
    if abs(earnings_yield) > 1.0 or abs(rf_current) > 1.0:
        raise ValueError(
            "Fatal Scale Error: earnings_yield and rf_current must be expressed "
            "as decimals (e.g., 0.032 for 3.2 %). "
            f"Received earnings_yield={earnings_yield}, rf_current={rf_current}."
        )

    # CAPE validity guard: CAPE must be strictly positive and non-NaN
    if pd.isna(cape_current) or cape_current <= 0.0:
        raise ValueError(
            f"cape_current must be a strictly positive finite number, "
            f"got {cape_current}."
        )

    # -----------------------------------------------------------------------
    # Core computation: ERP_impl = EarningsYield - rf
    # Equation: ERP_impl ≈ EarningsYield - r_f (Campbell-Shiller 1998)
    # -----------------------------------------------------------------------
    erp_implied: float = earnings_yield - rf_current

    # Point estimate: hat{mu}_5 = EarningsYield (total return proxy)
    estimate: float = earnings_yield

    # -----------------------------------------------------------------------
    # CAPE percentile computation: determines confidence via inverse-percentile
    # A higher CAPE percentile (more expensive) → lower confidence
    # -----------------------------------------------------------------------
    fallback_used: bool = False

    # Default percentile: neutral (0.5) used when history is unavailable
    cape_percentile: float = 0.5

    if cape_history is not None and not cape_history.empty:
        # Drop NaN values from the historical CAPE series before ranking
        clean_history: np.ndarray = cape_history.dropna().values

        if len(clean_history) > 0:
            # Count the number of historical CAPE observations at or below current
            n_below: int = int(np.sum(clean_history <= cape_current))

            # Empirical CDF percentile: fraction of history at or below current CAPE
            cape_percentile = float(n_below) / float(len(clean_history))

        else:
            # All values were NaN after dropna; fall back to neutral percentile
            fallback_used = True

    else:
        # No history provided; fall back to neutral percentile
        fallback_used = True

    if fallback_used:
        logger.warning(
            "Method 5: CAPE history is missing or entirely NaN. "
            "Defaulting to neutral percentile of %.2f.",
            cape_percentile,
        )

    # Confidence: c_5 = 1 - CAPE_percentile
    # High CAPE percentile (expensive) → low confidence in the earnings yield
    confidence: float = float(np.clip(1.0 - cape_percentile, 0.0, 1.0))

    # -----------------------------------------------------------------------
    # Rationale string: deterministic audit trail for the CMA Judge and board memo
    # -----------------------------------------------------------------------
    rationale: str = (
        f"Method 5 (Implied ERP / CAPE Earnings Yield): "
        f"EarningsYield = {earnings_yield * 100:.2f}%, "
        f"Rf = {rf_current * 100:.2f}%. "
        f"ERP_impl = {erp_implied * 100:.2f}%. "
        f"Estimate = {estimate * 100:.2f}%. "
        f"CAPE = {cape_current:.1f}x "
        f"(Percentile = {cape_percentile * 100:.1f}%). "
        f"Confidence = {confidence:.3f}."
    )

    # Emit a debug-level log entry for pipeline traceability
    logger.debug(
        "Executed Method 5: Estimate=%.4f, Confidence=%.4f", estimate, confidence
    )

    # -----------------------------------------------------------------------
    # Assemble and return the method-result dictionary (cma_methods.json schema)
    # -----------------------------------------------------------------------
    return {
        # Canonical method identifier consumed by Method 7 and the CMA Judge
        "method_id": 5,
        # Annualised expected return (earnings yield) expressed as a decimal
        "estimate": float(estimate),
        # CAPE-percentile-inverted confidence score in [0, 1]
        "confidence": float(confidence),
        # Component-level breakdown for audit and CMA Judge reasoning
        "breakdown": {
            # CAPE-based earnings yield (decimal)
            "earnings_yield": float(earnings_yield),
            # Current risk-free rate (decimal)
            "rf_current": float(rf_current),
            # Implied ERP = earnings_yield - rf (decimal)
            "erp_implied": float(erp_implied),
            # Current CAPE ratio (multiple, e.g., 25.0)
            "cape_current": float(cape_current),
            # Empirical CAPE percentile within historical distribution
            "cape_percentile": float(cape_percentile),
        },
        # Human-readable rationale for the CMA Judge skill and board memo
        "rationale": rationale,
    }


# -------------------------------------------------------------------------------------------------------------------------------
# Task 21, Step 4: Implement Method 6 (Survey) and Method 7 (Auto-Blend)
# -------------------------------------------------------------------------------------------------------------------------------

def execute_cma_method_6_survey(
    survey_expected_return: float,
    months_stale: float,
    decay_lambda: float,
) -> Dict[str, Any]:
    """
    Computes CMA Method 6: Survey / Analyst Consensus Expected Return.

    Implements the sixth candidate expected-return estimator in the seven-method
    CMA pipeline described in Ang, Azimbayev, and Kim (2026), Section 3.2.  The
    estimator passes through the survey consensus return directly as the point
    estimate and applies an exponential staleness decay to the confidence score,
    reflecting the diminishing informativeness of survey data as it ages.

    Equation (Task 21, Step 3)
    --------------------------
    Point estimate:

    .. math::

        \\hat{\\mu}_6 = \\text{survey\\_expected\\_return}

    Confidence decay:

    .. math::

        c_6 = e^{-\\lambda \\cdot \\text{months\\_stale}}

    where :math:`\\lambda` is the exponential decay rate sourced from
    ``STUDY_CONFIG["METHODOLOGY_PARAMS"]["CMA_METHOD_6_DECAY_LAMBDA"]``.
    A freshly published survey (``months_stale = 0``) yields
    :math:`c_6 = 1.0`; confidence decays monotonically toward zero as the
    survey ages.

    Pipeline Position
    -----------------
    * Upstream  : survey data loader (``df_survey_cma_raw``) in the
                  orchestrator; ``survey_expected_return`` column.
    * Downstream: ``execute_cma_method_7_auto_blend`` (Method 7 aggregator).
    * Output    : one element of the ``method_results`` list written to
                  ``cma_methods.json`` by the orchestrator.

    Parameters
    ----------
    survey_expected_return : float
        Analyst or institutional consensus expected return expressed as a
        **decimal** (e.g., 0.075 for 7.5 %).  Must satisfy
        ``|survey_expected_return| <= 1.0``.
    months_stale : float
        Number of months elapsed since the survey was published, measured
        relative to ``as_of_date``.  Must be a finite, non-NaN float.
        Negative values (future-dated surveys) are clamped to 0.0 internally
        with a warning.
    decay_lambda : float
        Exponential decay rate :math:`\\lambda` governing how rapidly
        confidence falls with survey age.  Sourced from
        ``STUDY_CONFIG["METHODOLOGY_PARAMS"]["CMA_METHOD_6_DECAY_LAMBDA"]``.
        A value of 0.1 implies that a 10-month-old survey retains
        :math:`e^{-1} \\approx 36.8\\%` of its original confidence.

    Returns
    -------
    Dict[str, Any]
        A method-result dictionary conforming to the ``cma_methods.json`` output
        contract.  Keys:

        ``method_id`` : int
            Always 6.
        ``estimate`` : float
            Survey consensus expected return expressed as a **decimal**.
        ``confidence`` : float
            Exponential staleness-decayed confidence score; lies in [0, 1].
        ``breakdown`` : Dict[str, Any]
            Component-level audit trail:

            * ``months_stale``  — raw staleness in months (may be negative).
            * ``decay_lambda``  — exponential decay rate used.

        ``rationale`` : str
            Human-readable one-line audit string for the CMA Judge skill.

    Raises
    ------
    ValueError
        If ``|survey_expected_return| > 1.0`` (scale guard) or if
        ``months_stale`` is NaN.
    """
    # -----------------------------------------------------------------------
    # Decimal scale guard: survey_expected_return must be in decimal form
    # -----------------------------------------------------------------------
    if abs(survey_expected_return) > 1.0:
        raise ValueError(
            f"Fatal Scale Error: survey_expected_return ({survey_expected_return}) "
            "exceeds the 1.0 absolute limit. Must be expressed as a decimal "
            "(e.g., 0.075 for 7.5 %)."
        )

    # NaN guard: months_stale must be a finite number for the decay computation
    if pd.isna(months_stale):
        raise ValueError(
            "months_stale evaluates to NaN. "
            "Provide a finite float representing months since survey publication."
        )

    # -----------------------------------------------------------------------
    # Point estimate: hat{mu}_6 = survey_expected_return (pass-through)
    # -----------------------------------------------------------------------
    estimate: float = survey_expected_return

    # -----------------------------------------------------------------------
    # Staleness clamping: future-dated surveys (months_stale < 0) are treated
    # as having zero staleness; the max(0, ...) guard prevents exp(-lambda * neg)
    # from inflating confidence above 1.0
    # -----------------------------------------------------------------------
    safe_stale: float = max(0.0, months_stale)

    # Confidence decay: c_6 = exp(-lambda * months_stale)
    # Equation: exponential decay of survey informativeness with age
    confidence: float = float(np.exp(-decay_lambda * safe_stale))

    # Clip to [0, 1] as a defensive guard against floating-point edge cases
    confidence = float(np.clip(confidence, 0.0, 1.0))

    # -----------------------------------------------------------------------
    # Rationale string: deterministic audit trail for the CMA Judge and board memo
    # -----------------------------------------------------------------------
    rationale: str = (
        f"Method 6 (Survey / Analyst Consensus): "
        f"Survey Return = {estimate * 100:.2f}%. "
        f"Data is {months_stale:.1f} months stale. "
        f"Exponential decay applied (lambda={decay_lambda:.4f}). "
        f"Confidence = {confidence:.3f}."
    )

    # Emit a debug-level log entry for pipeline traceability
    logger.debug(
        "Executed Method 6: Estimate=%.4f, Confidence=%.4f", estimate, confidence
    )

    # -----------------------------------------------------------------------
    # Assemble and return the method-result dictionary (cma_methods.json schema)
    # -----------------------------------------------------------------------
    return {
        # Canonical method identifier consumed by Method 7 and the CMA Judge
        "method_id": 6,
        # Survey consensus expected return expressed as a decimal
        "estimate": float(estimate),
        # Exponential staleness-decayed confidence score in [0, 1]
        "confidence": float(confidence),
        # Component-level breakdown for audit and CMA Judge reasoning
        "breakdown": {
            # Raw staleness in months (may be negative before clamping)
            "months_stale": float(months_stale),
            # Exponential decay rate used in the confidence computation
            "decay_lambda": float(decay_lambda),
        },
        # Human-readable rationale for the CMA Judge skill and board memo
        "rationale": rationale,
    }


def execute_cma_method_7_auto_blend(
    method_results: List[Dict[str, Any]],
    epsilon: float = 1e-8,
) -> Dict[str, Any]:
    """
    Computes CMA Method 7: Confidence-Weighted Auto-Blend of Methods 1–6.

    Implements the seventh and final candidate expected-return estimator in the
    CMA pipeline described in Ang, Azimbayev, and Kim (2026), Section 3.2.  The
    estimator aggregates the point estimates from Methods 1–6 using a
    confidence-weighted average, producing a robust prior that degrades
    gracefully when individual methods fail or return ``None`` estimates.

    Equation (Task 21, Step 3)
    --------------------------
    .. math::

        \\hat{\\mu}_7
            = \\frac{\\sum_{k=1}^{6} c_k \\, \\hat{\\mu}_k}
                    {\\sum_{k=1}^{6} c_k + \\epsilon}

    where :math:`c_k \\in [0, 1]` is the confidence score for method :math:`k`
    and :math:`\\epsilon = 10^{-8}` prevents zero-division when all confidences
    are zero.

    Methods with ``estimate = None`` (i.e., methods that failed and produced a
    stub result) are **excluded** from both the numerator and the denominator
    before any arithmetic is performed, implementing the Task 21 Step 3
    omission-prevention requirement.

    Confidence Scoring
    ------------------
    The auto-blend confidence is the self-weighted mean of the individual
    confidences, normalised by the same denominator:

    .. math::

        c_7 = \\text{clip}\\!\\left(
                  \\frac{\\sum_{k} c_k^2}{\\sum_{k} c_k + \\epsilon},
                  \\; 0, \\; 1
              \\right)

    This quantity is higher when the contributing methods agree (high, uniform
    confidences) and lower when the blend is dominated by a single high-confidence
    method surrounded by low-confidence peers.

    Pipeline Position
    -----------------
    * Upstream  : all six method-result dictionaries produced by Methods 1–6.
    * Downstream: CMA Judge skill (LLM-as-judge step in the AC Agent).
    * Output    : the seventh and final element of the ``method_results`` list
                  written to ``cma_methods.json`` by the orchestrator.

    Parameters
    ----------
    method_results : List[Dict[str, Any]]
        List of method-result dictionaries produced by Methods 1–6.  Each
        dictionary must contain at minimum the keys ``"method_id"``,
        ``"estimate"``, and ``"confidence"``.  Dictionaries with
        ``method_id`` outside ``{1, 2, 3, 4, 5, 6}`` are silently skipped.
        Dictionaries with ``estimate = None`` are excluded from the blend.
    epsilon : float, optional
        Small positive floor added to the confidence denominator to prevent
        zero-division when all valid method confidences are zero.  Default is
        1e-8.  Per Task 21 Step 3: "adding a numerical epsilon (epsilon = 1e-8)
        to the denominator."

    Returns
    -------
    Dict[str, Any]
        A method-result dictionary conforming to the ``cma_methods.json`` output
        contract.  Keys:

        ``method_id`` : int
            Always 7.
        ``estimate`` : float
            Confidence-weighted blend of valid method estimates expressed as a
            **decimal**.
        ``confidence`` : float
            Self-weighted mean confidence of the contributing methods; lies in
            [0, 1].
        ``breakdown`` : Dict[str, Any]
            Component-level audit trail:

            * ``n_valid_methods``    — number of methods included in the blend.
            * ``sum_confidences``    — sum of contributing confidence scores.
            * ``per_method_weights`` — normalised weight assigned to each method.

        ``rationale`` : str
            Human-readable one-line audit string for the CMA Judge skill.

    Raises
    ------
    ValueError
        If no valid methods remain after filtering out ``None`` estimates and
        out-of-range method IDs (i.e., the blend cannot be computed).
    """
    # -----------------------------------------------------------------------
    # Define the canonical set of method IDs eligible for blending (1 through 6)
    # Method 7 itself is excluded to prevent circular self-reference
    # -----------------------------------------------------------------------
    valid_ids: set = {1, 2, 3, 4, 5, 6}

    # -----------------------------------------------------------------------
    # Filter: retain only methods with a valid ID and a non-None estimate
    # Task 21 Step 3: "Ensure any method returning None (due to missing data)
    # is excluded from the Method 7 numerator and denominator."
    # -----------------------------------------------------------------------
    valid_methods: List[Dict[str, Any]] = []

    for m in method_results:
        # Skip any method whose ID is not in the canonical eligible set
        if m["method_id"] not in valid_ids:
            continue

        # Skip any method that produced a None estimate (stub / failure result)
        if m.get("estimate") is None:
            continue

        # Retain the method for blending
        valid_methods.append(m)

    # Guard: at least one valid method must remain for the blend to be meaningful
    if not valid_methods:
        raise ValueError(
            "Auto-Blend (Method 7) failed: no valid method estimates remain "
            "after filtering None estimates and out-of-range method IDs. "
            "At least one of Methods 1–6 must produce a non-None estimate."
        )

    # -----------------------------------------------------------------------
    # Extract estimates and confidences as aligned NumPy arrays for vectorised
    # arithmetic — avoids Python-level loops over the valid_methods list
    # -----------------------------------------------------------------------
    estimates: np.ndarray = np.array(
        [m["estimate"] for m in valid_methods], dtype=np.float64
    )

    # Confidence scores for each valid method
    confidences: np.ndarray = np.array(
        [m["confidence"] for m in valid_methods], dtype=np.float64
    )

    # -----------------------------------------------------------------------
    # Denominator: sum of confidences + epsilon
    # Equation: denominator = sum_k c_k + epsilon
    # The epsilon floor prevents zero-division when all confidences are zero
    # -----------------------------------------------------------------------
    sum_confidences: float = float(np.sum(confidences))

    # Denominator with epsilon floor for numerical stability
    denominator: float = sum_confidences + epsilon

    # -----------------------------------------------------------------------
    # Weighted estimate: hat{mu}_7 = sum(c_k * hat{mu}_k) / denominator
    # Equation: confidence-weighted average of valid method estimates
    # -----------------------------------------------------------------------
    estimate: float = float(np.sum(confidences * estimates) / denominator)

    # -----------------------------------------------------------------------
    # Self-weighted confidence: c_7 = clip(sum(c_k^2) / denominator, 0, 1)
    # Higher when contributing methods have uniformly high, consistent confidences
    # -----------------------------------------------------------------------
    confidence: float = float(np.sum(confidences * confidences) / denominator)

    # Clip to [0, 1] as a defensive guard against floating-point edge cases
    confidence = float(np.clip(confidence, 0.0, 1.0))

    # -----------------------------------------------------------------------
    # Normalised per-method weights for the breakdown audit trail
    # w_k = c_k / denominator (sums to approximately 1.0 modulo epsilon)
    # -----------------------------------------------------------------------
    normalized_weights: np.ndarray = confidences / denominator

    # Map each method ID to its normalised blend weight for the audit trail
    per_method_weights: Dict[str, float] = {
        f"method_{m['method_id']}": float(w)
        for m, w in zip(valid_methods, normalized_weights)
    }

    # -----------------------------------------------------------------------
    # Rationale string: deterministic audit trail for the CMA Judge and board memo
    # -----------------------------------------------------------------------
    n_valid: int = len(valid_methods)

    rationale: str = (
        f"Method 7 (Confidence-Weighted Auto-Blend): "
        f"Blended {n_valid} valid methods (IDs: "
        f"{[m['method_id'] for m in valid_methods]}). "
        f"Estimate = {estimate * 100:.2f}%. "
        f"Self-weighted Confidence = {confidence:.3f}. "
        f"Normalised weights: {per_method_weights}."
    )

    # Emit a debug-level log entry for pipeline traceability
    logger.debug(
        "Executed Method 7: Estimate=%.4f, Confidence=%.4f", estimate, confidence
    )

    # -----------------------------------------------------------------------
    # Assemble and return the method-result dictionary (cma_methods.json schema)
    # -----------------------------------------------------------------------
    return {
        # Canonical method identifier — always 7 for the auto-blend
        "method_id": 7,
        # Confidence-weighted blend of valid method estimates expressed as a decimal
        "estimate": float(estimate),
        # Self-weighted mean confidence of contributing methods in [0, 1]
        "confidence": float(confidence),
        # Component-level breakdown for audit and CMA Judge reasoning
        "breakdown": {
            # Number of valid (non-None) methods included in the blend
            "n_valid_methods": int(n_valid),
            # Raw sum of contributing confidence scores (before epsilon addition)
            "sum_confidences": float(sum_confidences),
            # Normalised weight assigned to each contributing method
            "per_method_weights": per_method_weights,
        },
        # Human-readable rationale for the CMA Judge skill and board memo
        "rationale": rationale,
    }


# -------------------------------------------------------------------------------------------------------------------------------
# Task 21, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_21_equity_cma_computation(
    asset_class: str,
    asset_index: int,
    as_of_date: str,
    historical_stats: Dict[str, Any],
    macro_view: Dict[str, Any],
    rf_current: float,
    rf_series: pd.Series,
    regime_history: pd.Series,
    sigma: np.ndarray,
    w_mkt: np.ndarray,
    df_fundamentals_raw: pd.DataFrame,
    df_survey_cma_raw: pd.DataFrame,
    config: Dict[str, Any],
) -> List[Dict[str, Any]]:
    """
    Orchestrates the sequential execution of Equity CMA candidate methods 1–7
    for a single equity asset class (Task 21).

    This function is the top-level script callable invoked by the Asset Class
    (AC) Agent tool-call layer (Task 18).  It chains the seven CMA method
    functions in the prescribed order, handles per-method failures gracefully
    via stub results, and returns a validated list of seven method-result
    dictionaries for serialisation to ``cma_methods.json``.

    Pipeline
    --------
    1. Extract and validate all required configuration parameters from
       ``STUDY_CONFIG`` — fail immediately on any missing key.
    2. Validate the ``monthly_returns_series`` and ``macro_view`` inputs.
    3. Execute Methods 1–7 sequentially, catching per-method ``ValueError``
       and ``IndexError`` exceptions and replacing failed methods with stub
       results (``estimate=None``, ``confidence=0.0``).
    4. Assert that exactly seven method-result dictionaries are returned with
       sequential IDs 1–7.

    Stub Behaviour
    --------------
    If a method raises an exception, a stub dictionary is inserted:

    .. code-block:: python

        {
            "method_id": k,
            "estimate": None,
            "confidence": 0.0,
            "breakdown": {"failed": True, "reason": "<exception message>"},
            "rationale": "Failed: <exception message>",
        }

    Method 7 (auto-blend) automatically excludes stub results (``estimate=None``)
    from its weighted average, implementing graceful degradation.

    Parameters
    ----------
    asset_class : str
        Canonical asset class label matching an entry in
        ``STUDY_CONFIG["IPS_GOVERNANCE"]["ASSET_UNIVERSE"]``
        (e.g., ``"US Large Cap"``).
    asset_index : int
        Zero-based index of ``asset_class`` within the 18-asset universe.
        Must satisfy ``0 <= asset_index < len(asset_universe)``.
    as_of_date : str
        ISO-8601 date string (e.g., ``"2026-03-31"``) representing the
        point-in-time cutoff for all data lookups.  All data filtered to
        ``<= as_of_date`` to prevent lookahead bias.
    historical_stats : Dict[str, Any]
        Dictionary produced by the historical analysis script (Task 18).
        Must contain the key ``"monthly_returns_series"`` mapping to a
        ``pd.Series`` with a ``pd.DatetimeIndex`` of decimal monthly returns.
    macro_view : Dict[str, Any]
        The validated macro_view dictionary produced by Task 20.  Must contain:

        * ``"regime"``     — str, one of the canonical regime labels.
        * ``"confidence"`` — float in [0, 1], macro-agent softmax confidence.

    rf_current : float
        Current annualised risk-free rate expressed as a **decimal**.
    rf_series : pd.Series
        Monthly risk-free rate series with a ``pd.DatetimeIndex``, expressed
        as **decimal** period rates.
    regime_history : pd.Series
        Monthly regime-label series with a ``pd.DatetimeIndex``, aligned to
        the same frequency as ``rf_series``.
    sigma : np.ndarray
        Annualised ``(18, 18)`` asset-class covariance matrix produced by the
        Covariance Agent.
    w_mkt : np.ndarray
        Market-capitalisation weight vector of shape ``(18,)`` summing to 1.0.
    df_fundamentals_raw : pd.DataFrame
        Multi-indexed fundamental data frame with a ``pd.MultiIndex`` of
        ``(date, ticker)``.  Must contain columns: ``"dividend_yield"``,
        ``"buyback_yield"``, ``"earnings_growth_forecast"``,
        ``"valuation_change_assumption"``, ``"earnings_yield"``,
        ``"cape_ratio"``.
    df_survey_cma_raw : pd.DataFrame
        Survey CMA data frame with a ``pd.DatetimeIndex`` and columns
        ``"asset_class"`` and ``"survey_expected_return"``.
    config : Dict[str, Any]
        The validated ``STUDY_CONFIG`` dictionary.  Required keys:

        * ``IPS_GOVERNANCE.ASSET_UNIVERSE``                          — List[str]
        * ``METHODOLOGY_PARAMS.CMA_METHOD_1_PERIODS_PER_YEAR``       — int
        * ``METHODOLOGY_PARAMS.CMA_METHOD_2_FALLBACK_PENALTY``       — float
        * ``METHODOLOGY_PARAMS.BL_EQUILIBRIUM_PARAMS.delta``         — float
        * ``METHODOLOGY_PARAMS.NOMINAL_GDP_GROWTH_PROXY``            — float
        * ``METHODOLOGY_PARAMS.CMA_METHOD_4_CONF_EXPLICIT``          — float
        * ``METHODOLOGY_PARAMS.CMA_METHOD_4_CONF_PROXY``             — float
        * ``METHODOLOGY_PARAMS.CMA_METHOD_6_DECAY_LAMBDA``           — float

    Returns
    -------
    List[Dict[str, Any]]
        An ordered list of exactly seven method-result dictionaries with
        sequential ``method_id`` values 1–7, conforming to the
        ``cma_methods.json`` output contract.  Failed methods are represented
        by stub dictionaries with ``estimate=None`` and ``confidence=0.0``.

    Raises
    ------
    KeyError
        If any required ``STUDY_CONFIG`` key is missing.
    IndexError
        If ``asset_index`` is out of bounds for the configured asset universe.
    TypeError
        If ``monthly_returns_series`` is not a ``pd.Series`` or does not carry
        a ``pd.DatetimeIndex``.
    ValueError
        If ``macro_view["regime"]`` is not a non-empty string or
        ``macro_view["confidence"]`` is outside [0, 1].
    AssertionError
        If the output list does not contain exactly seven dictionaries with
        sequential IDs 1–7.
    """
    logger.info(
        "Task 21: Starting Equity CMA computation for asset class '%s' "
        "(index=%d, as_of='%s').",
        asset_class,
        asset_index,
        as_of_date,
    )

    # -----------------------------------------------------------------------
    # Configuration extraction: pull all required methodology parameters from
    # STUDY_CONFIG in a single try/except block so that any missing key is
    # surfaced immediately with a precise error message identifying the exact
    # config path — fail-fast prevents silent misconfiguration propagation
    # -----------------------------------------------------------------------
    try:
        # Top-level methodology sub-dictionary containing all frozen parameters
        methodology: Dict[str, Any] = config["METHODOLOGY_PARAMS"]

        # Method 1: number of return periods per calendar year (12 for monthly)
        periods_per_year: int = methodology["CMA_METHOD_1_PERIODS_PER_YEAR"]

        # Method 2: multiplicative confidence penalty applied when the regime
        # fallback to the unconditional mean is triggered
        m2_fallback_penalty: float = methodology["CMA_METHOD_2_FALLBACK_PENALTY"]

        # Method 3: Black-Litterman risk-aversion coefficient delta (frozen at 2.5)
        bl_delta: float = methodology["BL_EQUILIBRIUM_PARAMS"]["delta"]

        # Method 4: nominal GDP growth proxy used when analyst forecasts are absent
        nominal_gdp_proxy: float = methodology["NOMINAL_GDP_GROWTH_PROXY"]

        # Method 4: confidence score applied when explicit analyst growth is available
        m4_conf_explicit: float = methodology["CMA_METHOD_4_CONF_EXPLICIT"]

        # Method 4: confidence score applied when the GDP proxy is used instead
        m4_conf_proxy: float = methodology["CMA_METHOD_4_CONF_PROXY"]

        # Method 6: exponential decay rate lambda for survey staleness penalty
        m6_decay_lambda: float = methodology["CMA_METHOD_6_DECAY_LAMBDA"]

        # IPS-defined 18-asset universe — used for asset_index bounds checking
        asset_universe: List[str] = config["IPS_GOVERNANCE"]["ASSET_UNIVERSE"]

    except KeyError as exc:
        raise KeyError(
            f"Fatal: STUDY_CONFIG is missing a required key for Task 21: {exc}. "
            "Ensure METHODOLOGY_PARAMS contains CMA_METHOD_1_PERIODS_PER_YEAR, "
            "CMA_METHOD_2_FALLBACK_PENALTY, BL_EQUILIBRIUM_PARAMS.delta, "
            "NOMINAL_GDP_GROWTH_PROXY, CMA_METHOD_4_CONF_EXPLICIT, "
            "CMA_METHOD_4_CONF_PROXY, CMA_METHOD_6_DECAY_LAMBDA, and "
            "IPS_GOVERNANCE.ASSET_UNIVERSE."
        ) from exc

    # -----------------------------------------------------------------------
    # Asset index bounds guard: asset_index must be a valid zero-based index
    # into the IPS-defined 18-asset universe
    # -----------------------------------------------------------------------
    if not (0 <= asset_index < len(asset_universe)):
        raise IndexError(
            f"asset_index {asset_index} is out of bounds for the configured "
            f"asset universe of length {len(asset_universe)}."
        )

    # Consistency warning: asset_class string should match the universe mapping
    # at asset_index; a mismatch indicates a caller-side wiring error
    if asset_universe[asset_index] != asset_class:
        logger.warning(
            "Task 21: asset_class '%s' does not match asset_universe[%d]='%s'. "
            "Verify caller-side asset_index assignment.",
            asset_class,
            asset_index,
            asset_universe[asset_index],
        )

    # -----------------------------------------------------------------------
    # monthly_returns_series extraction and type validation
    # The series must be a pd.Series with a DatetimeIndex; any other type
    # indicates a malformed historical_stats dictionary from the data-fetch script
    # -----------------------------------------------------------------------
    monthly_returns: pd.Series = historical_stats["monthly_returns_series"]

    # Enforce pd.Series type for the monthly returns input
    if not isinstance(monthly_returns, pd.Series):
        raise TypeError(
            "historical_stats['monthly_returns_series'] must be a pd.Series, "
            f"got {type(monthly_returns).__name__}."
        )

    # Enforce DatetimeIndex so that inner-join alignment in Methods 1 and 2
    # is temporally meaningful and not silently misaligned
    if not isinstance(monthly_returns.index, pd.DatetimeIndex):
        raise TypeError(
            "historical_stats['monthly_returns_series'] must carry a "
            "pd.DatetimeIndex for temporal alignment."
        )

    # -----------------------------------------------------------------------
    # macro_view validation: regime must be a non-empty string and confidence
    # must lie in [0, 1] — these are consumed by Methods 2 and the CMA Judge
    # -----------------------------------------------------------------------
    current_regime: str = macro_view["regime"]
    regime_confidence: float = macro_view["confidence"]

    # Regime label must be a non-empty string drawn from the canonical set
    if not isinstance(current_regime, str) or not current_regime:
        raise ValueError(
            "macro_view['regime'] must be a non-empty string "
            f"(e.g., 'Late-Cycle'), got: {current_regime!r}."
        )

    # Regime confidence must be a valid probability in [0, 1]
    if not (0.0 <= regime_confidence <= 1.0):
        raise ValueError(
            f"macro_view['confidence'] must lie in [0.0, 1.0], "
            f"got {regime_confidence}."
        )

    # -----------------------------------------------------------------------
    # Initialise the method_results accumulator list; will hold exactly 7 dicts
    # -----------------------------------------------------------------------
    method_results: List[Dict[str, Any]] = []

    # -----------------------------------------------------------------------
    # Stub factory: produces a failure-sentinel dictionary for any method that
    # raises an exception, preserving the sequential ID ordering required by
    # the post-hoc assertion and the cma_methods.json schema
    # -----------------------------------------------------------------------
    def _make_stub(method_id: int, rationale_msg: str) -> Dict[str, Any]:
        """
        Constructs a failure-sentinel method-result dictionary.

        Used internally by the orchestrator to represent a method that raised
        an exception during execution.  The ``estimate=None`` sentinel causes
        Method 7 (auto-blend) to exclude this entry from its weighted average,
        implementing graceful degradation per Task 21 Step 3.

        Parameters
        ----------
        method_id : int
            The canonical method identifier (1–7) of the failed method.
        rationale_msg : str
            The exception message string, preserved verbatim for audit trail.

        Returns
        -------
        Dict[str, Any]
            A stub method-result dictionary with ``estimate=None`` and
            ``confidence=0.0``.
        """
        # Return a sentinel dict that signals failure to downstream consumers
        return {
            # Canonical method identifier — preserved for sequential ID ordering
            "method_id": method_id,
            # None estimate signals exclusion from Method 7 auto-blend
            "estimate": None,
            # Zero confidence ensures this stub contributes nothing if included
            "confidence": 0.0,
            # Breakdown records the failure reason for the audit trail
            "breakdown": {
                "failed": True,
                "reason": rationale_msg,
            },
            # Human-readable rationale for the CMA Judge and board memo
            "rationale": f"Failed: {rationale_msg}",
        }

    # -----------------------------------------------------------------------
    # Method 1: Historical ERP + Current Risk-Free Rate
    # Equation: hat{mu}_1 = r_f + ERP_hist
    # -----------------------------------------------------------------------
    try:
        # Execute Method 1 using the full aligned history of returns and rf
        m1_result: Dict[str, Any] = execute_cma_method_1_historical_erp(
            monthly_returns=monthly_returns,
            rf_series=rf_series,
            rf_current=rf_current,
            periods_per_year=periods_per_year,
        )
        # Append the successful result to the accumulator
        method_results.append(m1_result)

    except ValueError as e:
        # Log the failure and append a stub to preserve sequential ordering
        logger.error("Task 21 Method 1 failed for '%s': %s", asset_class, e)
        method_results.append(_make_stub(1, str(e)))

    # -----------------------------------------------------------------------
    # Method 2: Regime-Adjusted ERP + Current Risk-Free Rate
    # Equation: hat{mu}_2 = r_f + ERP(regime)
    # -----------------------------------------------------------------------
    try:
        # Execute Method 2 conditioning on the current macro regime label
        m2_result: Dict[str, Any] = execute_cma_method_2_regime_adjusted(
            monthly_returns=monthly_returns,
            rf_series=rf_series,
            regime_history=regime_history,
            current_regime=current_regime,
            regime_confidence=regime_confidence,
            rf_current=rf_current,
            periods_per_year=periods_per_year,
            fallback_penalty=m2_fallback_penalty,
        )
        # Append the successful result to the accumulator
        method_results.append(m2_result)

    except ValueError as e:
        # Log the failure and append a stub to preserve sequential ordering
        logger.error("Task 21 Method 2 failed for '%s': %s", asset_class, e)
        method_results.append(_make_stub(2, str(e)))

    # -----------------------------------------------------------------------
    # Method 3: Black-Litterman Equilibrium Implied Return
    # Equation: pi = delta * Sigma * w_mkt; hat{mu}_3 = pi[asset_index]
    # -----------------------------------------------------------------------
    try:
        # Execute Method 3 using the covariance matrix and market-cap weights
        m3_result: Dict[str, Any] = execute_cma_method_3_bl_equilibrium(
            asset_index=asset_index,
            sigma=sigma,
            w_mkt=w_mkt,
            bl_delta=bl_delta,
        )
        # Append the successful result to the accumulator
        method_results.append(m3_result)

    except (ValueError, IndexError) as e:
        # Log the failure and append a stub to preserve sequential ordering
        logger.error("Task 21 Method 3 failed for '%s': %s", asset_class, e)
        method_results.append(_make_stub(3, str(e)))

    # -----------------------------------------------------------------------
    # Fundamental data extraction block: shared pre-processing for Methods 4 & 5
    # Both methods consume the same row of df_fundamentals_raw; extracting once
    # avoids redundant I/O and ensures both methods see identical input data
    # -----------------------------------------------------------------------
    try:
        # Guard: df_fundamentals_raw must carry a MultiIndex of (date, ticker)
        # so that xs() can slice by ticker without ambiguity
        if not isinstance(df_fundamentals_raw.index, pd.MultiIndex):
            raise TypeError(
                "df_fundamentals_raw must carry a pd.MultiIndex of (date, ticker). "
                f"Got index type: {type(df_fundamentals_raw.index).__name__}."
            )

        # Cross-section slice: extract all rows for the target asset class ticker
        df_fund_asset: pd.DataFrame = df_fundamentals_raw.xs(
            asset_class, level="ticker"
        )

        # Strip timezone information if present to ensure consistent Timestamp
        # comparison with the tz-naive as_of_date cutoff
        if df_fund_asset.index.tz is not None:
            df_fund_asset.index = df_fund_asset.index.tz_convert(None)

        # Point-in-time filter: retain only observations on or before as_of_date
        # to prevent lookahead bias in the fundamental building-block inputs
        df_fund_asset = df_fund_asset.loc[
            df_fund_asset.index <= pd.Timestamp(as_of_date)
        ]

        # Guard: at least one row must remain after the point-in-time filter
        if df_fund_asset.empty:
            raise ValueError(
                f"Fundamentals data is empty for '{asset_class}' "
                f"up to as_of_date='{as_of_date}'. "
                "Verify the data-fetch script and ticker mapping."
            )

        # Extract the most recent (latest) row of fundamental data
        latest_fundamentals: pd.Series = df_fund_asset.iloc[-1]

        # Dividend yield: y_div in the Grinold-Kroner building-block equation
        div_yield: float = float(latest_fundamentals.at["dividend_yield"])

        # Net buyback yield: y_buyback in the Grinold-Kroner equation
        buyback_yield: float = float(latest_fundamentals.at["buyback_yield"])

        # Earnings growth forecast: g in the Grinold-Kroner equation
        # If the analyst forecast is NaN, fall back to the nominal GDP proxy
        growth_raw = latest_fundamentals.at["earnings_growth_forecast"]
        growth_is_proxy: bool = bool(pd.isna(growth_raw))

        # Use the GDP proxy when the analyst forecast is missing; apply penalty
        earnings_growth: float = (
            float(nominal_gdp_proxy) if growth_is_proxy else float(growth_raw)
        )

        # Valuation change assumption: Delta_v in the Grinold-Kroner equation
        # Default to 0.0 (no valuation change assumed) if the field is NaN
        val_change_raw = latest_fundamentals.at["valuation_change_assumption"]
        val_change: float = 0.0 if pd.isna(val_change_raw) else float(val_change_raw)

        # Earnings yield: EarningsYield = 1 / CAPE for Method 5
        earnings_yield: float = float(latest_fundamentals.at["earnings_yield"])

        # Current CAPE ratio: used for Method 5 confidence percentile computation
        cape_current: float = float(latest_fundamentals.at["cape_ratio"])

        # Historical CAPE series: used to compute the current CAPE's percentile rank
        # within its own distribution for Method 5 confidence scoring
        cape_history: Optional[pd.Series] = (
            df_fund_asset["cape_ratio"]
            if "cape_ratio" in df_fund_asset.columns
            else None
        )

    except Exception as e:
        # Log the extraction failure; Methods 4 and 5 will both produce stubs
        logger.warning(
            "Task 21: Failed to extract fundamental data for '%s': %s. "
            "Methods 4 and 5 will produce stub results.",
            asset_class,
            e,
        )
        # Set all fundamental scalars to NaN so the per-method guards below
        # detect the failure and produce stubs rather than silently using 0.0
        div_yield = buyback_yield = earnings_growth = val_change = np.nan
        earnings_yield = cape_current = np.nan
        growth_is_proxy = True
        cape_history = None

    # -----------------------------------------------------------------------
    # Method 4: Inverse Gordon / Grinold-Kroner Building-Block Model
    # Equation: hat{mu}_4 = y_div + y_buyback + g + Delta_v
    # -----------------------------------------------------------------------
    try:
        # Guard: dividend and buyback yields must be non-NaN before calling Method 4
        # NaN inputs would silently propagate through the arithmetic sum
        if pd.isna(div_yield) or pd.isna(buyback_yield):
            raise ValueError(
                "Missing dividend_yield or buyback_yield in fundamentals data. "
                "Cannot compute Method 4 (Inverse Gordon)."
            )

        # Execute Method 4 with the extracted fundamental building-block inputs
        m4_result: Dict[str, Any] = execute_cma_method_4_inverse_gordon(
            dividend_yield=div_yield,
            buyback_yield=buyback_yield,
            earnings_growth=earnings_growth,
            valuation_change=val_change,
            conf_explicit=m4_conf_explicit,
            conf_proxy=m4_conf_proxy,
            growth_is_proxy=growth_is_proxy,
        )
        # Append the successful result to the accumulator
        method_results.append(m4_result)

    except ValueError as e:
        # Log the failure and append a stub to preserve sequential ordering
        logger.error("Task 21 Method 4 failed for '%s': %s", asset_class, e)
        method_results.append(_make_stub(4, str(e)))

    # -----------------------------------------------------------------------
    # Method 5: Implied ERP Proxy via Campbell-Shiller CAPE Earnings Yield
    # Equation: ERP_impl ≈ EarningsYield - r_f; hat{mu}_5 = EarningsYield
    # -----------------------------------------------------------------------
    try:
        # Guard: earnings_yield and cape_current must be non-NaN before calling
        # Method 5; NaN inputs would produce a nonsensical confidence percentile
        if pd.isna(earnings_yield) or pd.isna(cape_current):
            raise ValueError(
                "Missing earnings_yield or cape_ratio in fundamentals data. "
                "Cannot compute Method 5 (Implied ERP)."
            )

        # Execute Method 5 with the CAPE-based earnings yield and history
        m5_result: Dict[str, Any] = execute_cma_method_5_implied_erp(
            earnings_yield=earnings_yield,
            rf_current=rf_current,
            cape_current=cape_current,
            cape_history=cape_history,
        )
        # Append the successful result to the accumulator
        method_results.append(m5_result)

    except ValueError as e:
        # Log the failure and append a stub to preserve sequential ordering
        logger.error("Task 21 Method 5 failed for '%s': %s", asset_class, e)
        method_results.append(_make_stub(5, str(e)))

    # -----------------------------------------------------------------------
    # Method 6: Survey / Analyst Consensus Expected Return
    # Equation: hat{mu}_6 = survey_expected_return; c_6 = exp(-lambda * months_stale)
    # -----------------------------------------------------------------------
    try:
        # Filter the survey data frame to rows matching the target asset class
        df_survey_asset: pd.DataFrame = df_survey_cma_raw[
            df_survey_cma_raw["asset_class"] == asset_class
        ].copy()

        # Strip timezone information if present to ensure consistent Timestamp
        # comparison with the tz-naive as_of_date cutoff
        if df_survey_asset.index.tz is not None:
            df_survey_asset.index = df_survey_asset.index.tz_convert(None)

        # Point-in-time filter: retain only survey observations on or before as_of_date
        df_survey_asset = df_survey_asset.loc[
            df_survey_asset.index <= pd.Timestamp(as_of_date)
        ]

        # Guard: at least one survey row must remain after the point-in-time filter
        if df_survey_asset.empty:
            raise ValueError(
                f"No survey data available for '{asset_class}' "
                f"by as_of_date='{as_of_date}'. "
                "Verify the survey data source and asset_class label mapping."
            )

        # Extract the most recent survey expected return (decimal)
        survey_return: float = float(df_survey_asset.iloc[-1]["survey_expected_return"])

        # Record the publication date of the most recent survey observation
        survey_date: pd.Timestamp = df_survey_asset.index[-1]

        # Compute staleness in calendar days relative to as_of_date
        days_stale: float = float(
            (pd.Timestamp(as_of_date) - survey_date).days
        )

        # Warn if the survey is future-dated (days_stale < 0); the method
        # function clamps this to 0.0 internally via max(0, months_stale)
        if days_stale < 0.0:
            logger.warning(
                "Task 21 Method 6: Survey for '%s' is future-dated "
                "(days_stale=%.1f). Staleness will be clamped to 0.0.",
                asset_class,
                days_stale,
            )

        # Convert staleness from calendar days to months using the standard
        # average days-per-month conversion factor (365.25 / 12 ≈ 30.44)
        months_stale: float = days_stale / 30.44

        # Execute Method 6 with the survey return and computed staleness
        m6_result: Dict[str, Any] = execute_cma_method_6_survey(
            survey_expected_return=survey_return,
            months_stale=months_stale,
            decay_lambda=m6_decay_lambda,
        )
        # Append the successful result to the accumulator
        method_results.append(m6_result)

    except ValueError as e:
        # Log the failure and append a stub to preserve sequential ordering
        logger.error("Task 21 Method 6 failed for '%s': %s", asset_class, e)
        method_results.append(_make_stub(6, str(e)))

    # -----------------------------------------------------------------------
    # Method 7: Confidence-Weighted Auto-Blend of Methods 1–6
    # Equation: hat{mu}_7 = sum(c_k * hat{mu}_k) / (sum(c_k) + epsilon)
    # Stub results (estimate=None) are automatically excluded by Method 7
    # -----------------------------------------------------------------------
    try:
        # Execute Method 7 passing the full method_results list including any stubs
        m7_result: Dict[str, Any] = execute_cma_method_7_auto_blend(
            method_results=method_results,
        )
        # Append the successful auto-blend result to the accumulator
        method_results.append(m7_result)

    except ValueError as e:
        # Log the failure; a stub for Method 7 means no blend was possible
        # (all upstream methods failed), which is a critical pipeline failure
        logger.error(
            "Task 21 Method 7 (Auto-Blend) failed for '%s': %s. "
            "All upstream methods may have produced stub results.",
            asset_class,
            e,
        )
        method_results.append(_make_stub(7, str(e)))

    # -----------------------------------------------------------------------
    # Post-hoc output schema assertions: fail-closed validation ensuring the
    # downstream CMA Judge and cma_methods.json serialiser receive exactly the
    # structure they expect — any violation indicates an orchestration bug
    # -----------------------------------------------------------------------

    # Assert that exactly seven method-result dictionaries were produced
    assert len(method_results) == 7, (
        f"Task 21 output schema violation: expected exactly 7 method-result "
        f"dictionaries, got {len(method_results)}."
    )

    # Extract the method IDs from the output list for sequential ordering check
    ids: List[int] = [m["method_id"] for m in method_results]

    # Assert that method IDs are strictly sequential from 1 to 7 in order
    assert tuple(ids) == (1, 2, 3, 4, 5, 6, 7), (
        f"Task 21 output schema violation: method IDs must be strictly ordered "
        f"(1, 2, 3, 4, 5, 6, 7). Found: {ids}."
    )

    logger.info(
        "Task 21 completed successfully for asset class '%s'. "
        "Seven CMA method results produced.",
        asset_class,
    )

    # Return the validated list of seven method-result dictionaries
    return method_results
````

````python
# Task 22 — Execute CMA Judge (LLM-as-judge) exactly as specified (equity)

# ==============================================================================
# Task 22: Execute CMA Judge (LLM-as-judge) exactly as specified (equity)
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 22, Step 1: Provide to the LLM the required inputs and initialize the ReAct loop
# -------------------------------------------------------------------------------------------------------------------------------

def compile_cma_judge_prompt(
    prompt_templates: Dict[str, Any],
    asset_class: str
) -> Tuple[str, str]:
    """
    Retrieves the verbatim Exhibit 4 prompt for the CMA Judge and constructs
    the initial user message enforcing the mandatory tool sequence.

    Parameters
    ----------
    prompt_templates : Dict[str, Any]
        The dictionary containing all prompt templates from the study configuration.
    asset_class : str
        The canonical name of the asset class being evaluated.

    Returns
    -------
    Tuple[str, str]
        A tuple containing:
        1. The verbatim developer prompt string.
        2. The strict user instruction string.

    Raises
    ------
    KeyError
        If the required verbatim template is missing.
    ValueError
        If the template provenance is not strictly 'VERBATIM'.
    """
    template_key: str = "CMA_JUDGE_EXHIBIT4_VERBATIM"

    if template_key not in prompt_templates:
        raise KeyError(f"Prompt template '{template_key}' not found in configuration.")

    template_dict: Dict[str, Any] = prompt_templates[template_key]
    raw_content: str = template_dict.get("content", "")
    provenance: str = template_dict.get("provenance", "")

    # Strict provenance check: The CMA Judge prompt must be cryptographically identical to Exhibit 4
    if not provenance.startswith("VERBATIM"):
        raise ValueError(
            f"Fatal Methodological Error: Template '{template_key}' must have a provenance "
            f"starting with 'VERBATIM', got '{provenance}'. The CMA Judge logic cannot be mutated."
        )

    # Construct the strict user instruction enforcing the ReAct loop sequence
    user_instruction: str = (
        f"Begin CMA Judge evaluation for asset class '{asset_class}'.\n"
        f"NON-NEGOTIABLE: You must execute the following tools in this exact sequence:\n"
        f"1. load_cma_methods_json\n"
        f"2. classify_dispersion\n"
        f"3. check_regime_logic\n"
        f"4. check_valuation_thresholds\n"
        f"5. check_signal_alignment\n"
        f"6. enforce_range_constraint\n"
        f"7. write_cma_json\n"
        f"Do not perform arithmetic. Rely entirely on the tool outputs."
    )

    logger.debug(f"Successfully compiled CMA Judge prompt and user instructions for '{asset_class}'.")
    return raw_content, user_instruction

# -------------------------------------------------------------------------------------------------------------------------------
# Task 22, Step 2: Implement and register the Exhibit 4 algorithm as script tools
# -------------------------------------------------------------------------------------------------------------------------------

def register_cma_judge_tools(
    asset_class: str,
    artifact_dir: Path,
    macro_view: Dict[str, Any],
    signals: Dict[str, Any],
    actual_estimates: List[float]
) -> Tuple[Dict[str, Callable], List[Dict[str, Any]]]:
    """
    Constructs the tool registry for the CMA Judge, binding the deterministic
    Python callables to their JSON Schema definitions. Injects state variables
    via closures to prevent LLM tampering.

    Parameters
    ----------
    asset_class : str
        The canonical name of the asset class.
    artifact_dir : Path
        The base directory where output artifacts will be written.
    macro_view : Dict[str, Any]
        The parsed macro-view.json dictionary.
    signals : Dict[str, Any]
        The parsed signals.json dictionary for the asset class.
    actual_estimates : List[float]
        The true list of method estimates loaded from disk by the orchestrator.
        Used to securely bind the range constraint tool.

    Returns
    -------
    Tuple[Dict[str, Callable], List[Dict[str, Any]]]
        A tuple containing the callable registry and the JSON Schema array.
    """
    callable_registry: Dict[str, Callable] = {}
    tool_schemas: List[Dict[str, Any]] = []

    # 1. load_cma_methods_json
    callable_registry["load_cma_methods_json"] = partial(
        load_cma_methods_json,
        asset_class=asset_class,
        artifact_dir=artifact_dir
    )
    tool_schemas.append({
        "type": "function",
        "function": {
            "name": "load_cma_methods_json",
            "description": "Load and parse cma_methods.json for the given asset class.",
            "parameters": {"type": "object", "properties": {}, "required": []}
        }
    })

    # 2. classify_dispersion
    callable_registry["classify_dispersion"] = classify_dispersion
    tool_schemas.append({
        "type": "function",
        "function": {
            "name": "classify_dispersion",
            "description": "Classify the dispersion of CMA method estimates (tight, moderate, wide).",
            "parameters": {
                "type": "object",
                "properties": {
                    "estimates": {
                        "type": "array",
                        "items": {"type": "number"},
                        "description": "List of method estimates extracted from load_cma_methods_json."
                    }
                },
                "required": ["estimates"]
            }
        }
    })

    # 3. check_regime_logic
    # Inject macro_view to prevent LLM hallucination of the regime
    callable_registry["check_regime_logic"] = partial(
        check_regime_logic,
        macro_view=macro_view
    )
    tool_schemas.append({
        "type": "function",
        "function": {
            "name": "check_regime_logic",
            "description": "Apply the regime-conditional method selection hint from Exhibit 4 Step 2.",
            "parameters": {
                "type": "object",
                "properties": {
                    "dispersion": {
                        "type": "object",
                        "description": "The output dictionary from classify_dispersion."
                    }
                },
                "required": ["dispersion"]
            }
        }
    })

    # 4. check_valuation_thresholds
    # Inject signals to prevent LLM hallucination of the P/E ratio
    callable_registry["check_valuation_thresholds"] = partial(
        check_valuation_thresholds,
        signals=signals
    )
    tool_schemas.append({
        "type": "function",
        "function": {
            "name": "check_valuation_thresholds",
            "description": "Apply the valuation threshold check from Exhibit 4 Step 3.",
            "parameters": {"type": "object", "properties": {}, "required": []}
        }
    })

    # 5. check_signal_alignment
    # Inject signals to prevent LLM hallucination of technical indicators
    callable_registry["check_signal_alignment"] = partial(
        check_signal_alignment,
        signals=signals
    )
    tool_schemas.append({
        "type": "function",
        "function": {
            "name": "check_signal_alignment",
            "description": "Compute the alignment between technical signals and the proposed CMA estimate.",
            "parameters": {
                "type": "object",
                "properties": {
                    "final_estimate": {
                        "type": "number",
                        "description": "The proposed final CMA estimate in decimal form."
                    }
                },
                "required": ["final_estimate"]
            }
        }
    })

    # 6. enforce_range_constraint
    # Inject the actual estimates loaded from disk to prevent the LLM from passing a fake array
    callable_registry["enforce_range_constraint"] = partial(
        enforce_range_constraint,
        estimates=actual_estimates
    )
    tool_schemas.append({
        "type": "function",
        "function": {
            "name": "enforce_range_constraint",
            "description": "Enforce the hard range constraint from Exhibit 4 (Layer 1 gate).",
            "parameters": {
                "type": "object",
                "properties": {
                    "final_estimate": {
                        "type": "number",
                        "description": "The proposed final CMA estimate in decimal form."
                    }
                },
                "required": ["final_estimate"]
            }
        }
    })

    # 7. write_cma_json
    callable_registry["write_cma_json"] = partial(
        write_cma_json,
        asset_class=asset_class,
        artifact_dir=artifact_dir
    )
    tool_schemas.append({
        "type": "function",
        "function": {
            "name": "write_cma_json",
            "description": "Serialise and persist the final CMA estimate to cma.json.",
            "parameters": {
                "type": "object",
                "properties": {
                    "final_estimate": {"type": "number"},
                    "method_weights": {"type": "object"},
                    "rationale": {"type": "string"},
                    "method_range": {
                        "type": "object",
                        "properties": {
                            "min": {"type": "number"},
                            "max": {"type": "number"}
                        },
                        "required": ["min", "max"]
                    }
                },
                "required": ["final_estimate", "method_weights", "rationale", "method_range"]
            }
        }
    })

    logger.debug(f"Successfully registered 7 CMA Judge tools for '{asset_class}'.")
    return callable_registry, tool_schemas

# -------------------------------------------------------------------------------------------------------------------------------
# Task 22, Step 3: Enforce the hard range constraint with a dual-layer gate (ReAct Loop)
# -------------------------------------------------------------------------------------------------------------------------------

def execute_cma_judge_react_loop(
    asset_class: str,
    client: Any, # OpenAI Client
    llm_config: AgentLLMConfig,
    developer_prompt: str,
    user_instruction: str,
    callable_registry: Dict[str, Callable],
    tool_schemas: List[Dict[str, Any]],
    mandatory_sequence: List[str],
    max_iterations: int = 20
) -> str:
    """
    Executes the ReAct loop for the CMA Judge. Enforces the strict tool-call
    sequence and the Layer 1 range constraint gate.

    Parameters
    ----------
    asset_class : str
        The canonical name of the asset class.
    client : Any
        The authenticated OpenAI client.
    llm_config : AgentLLMConfig
        The strictly typed configuration for the agent.
    developer_prompt : str
        The verbatim Exhibit 4 prompt.
    user_instruction : str
        The strict instruction enforcing the tool sequence.
    callable_registry : Dict[str, Callable]
        The dictionary mapping tool names to Python functions.
    tool_schemas : List[Dict[str, Any]]
        The JSON Schema array of available tools.
    mandatory_sequence : List[str]
        The exact order in which tools must be called.
    max_iterations : int, optional
        The hard iteration ceiling. Defaults to 20.

    Returns
    -------
    str
        The path to the written cma.json artifact.

    Raises
    ------
    RuntimeError
        If the iteration ceiling is hit without clean termination.
    """
    messages: List[Dict[str, str]] = [
        {"role": "developer", "content": developer_prompt},
        {"role": "user", "content": user_instruction}
    ]

    # State tracker for the mandatory sequence
    sequence_index: int = 0
    cma_json_path: str = ""

    iteration: int = 0
    while iteration < max_iterations:
        iteration += 1
        logger.debug(f"CMA Judge ({asset_class}) ReAct Loop Iteration {iteration}/{max_iterations}")

        response: ParsedLLMResponse = invoke_and_extract_agent_response(
            client=client,
            config=llm_config,
            messages=messages,
            tools=tool_schemas
        )

        assistant_msg: Dict[str, Any] = {"role": "assistant"}
        if response.content:
            assistant_msg["content"] = response.content
        if response.tool_calls:
            assistant_msg["tool_calls"] = response.tool_calls
        messages.append(assistant_msg)

        if response.tool_calls:
            for tool_call in response.tool_calls:
                tool_name: str = tool_call.function.name

                # Sequence Enforcement Gate
                expected_tool: str = mandatory_sequence[sequence_index] if sequence_index < len(mandatory_sequence) else "None"
                if tool_name != expected_tool:
                    error_msg: str = f"SEQUENCE VIOLATION: Expected tool '{expected_tool}', but you called '{tool_name}'. You must follow the mandatory sequence."
                    logger.warning(f"CMA Judge ({asset_class}) - {error_msg}")
                    messages.append({
                        "role": "tool",
                        "tool_call_id": tool_call.id,
                        "name": tool_name,
                        "content": error_msg
                    })
                    continue # Skip execution and force LLM to retry

                try:
                    tool_args: Dict[str, Any] = json.loads(tool_call.function.arguments)
                except json.JSONDecodeError as exc:
                    error_msg = f"JSONDecodeError: Failed to parse arguments for tool {tool_name}. Details: {exc}"
                    logger.error(f"CMA Judge ({asset_class}) - {error_msg}")
                    messages.append({
                        "role": "tool",
                        "tool_call_id": tool_call.id,
                        "name": tool_name,
                        "content": error_msg
                    })
                    continue

                if tool_name in callable_registry:
                    try:
                        result: Any = callable_registry[tool_name](**tool_args)

                        # Track termination condition
                        if tool_name == "write_cma_json":
                            cma_json_path = str(result)

                        # Advance the sequence tracker
                        sequence_index += 1

                        result_str: str = json.dumps(result)
                    except Exception as exc:
                        result_str = f"Error executing {tool_name}: {str(exc)}"
                        logger.error(f"CMA Judge ({asset_class}) - {result_str}")
                else:
                    result_str = f"Error: Tool {tool_name} not found in registry."

                messages.append({
                    "role": "tool",
                    "tool_call_id": tool_call.id,
                    "name": tool_name,
                    "content": result_str
                })

        # Check termination condition: all tools in sequence called
        if sequence_index >= len(mandatory_sequence):
            logger.info(f"CMA Judge ({asset_class}) ReAct loop terminated cleanly.")
            return cma_json_path

        # Prompt to continue if stalled
        if not response.tool_calls and sequence_index < len(mandatory_sequence):
            expected_tool = mandatory_sequence[sequence_index]
            messages.append({
                "role": "user",
                "content": f"You have not completed the workflow. You must call '{expected_tool}' next."
            })

    raise RuntimeError(f"CMA Judge ({asset_class}) ReAct loop hit iteration ceiling ({max_iterations}).")

# -------------------------------------------------------------------------------------------------------------------------------
# Task 22, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_22_cma_judge(
    asset_class: str,
    config: Dict[str, Any],
    run_output_dir: Path,
    client: Any, # OpenAI Client
    validate_artifact_fn: Callable[[Path, str], bool]
) -> str:
    """
    Orchestrates the CMA Judge package for a single equity asset class.
    Executes the ReAct loop to evaluate method dispersion and applies the
    Layer 2 post-hoc range constraint verification.

    CRITICAL REMEDIATION: Replaces the fragile `min()` and `max()` list reductions
    with robust `np.nanmin()` and `np.nanmax()` operations on a strictly typed
    NumPy array. This prevents `TypeError` crashes if upstream methods failed
    and produced `None` or `NaN` stubs.

    Parameters
    ----------
    asset_class : str
        The canonical name of the equity asset class.
    config : Dict[str, Any]
        The validated study configuration dictionary.
    run_output_dir : Path
        The root directory for the current run's artifacts.
    client : Any
        The authenticated OpenAI client.
    validate_artifact_fn : Callable[[Path, str], bool]
        The bound validation function from Task 16.

    Returns
    -------
    str
        Path to the generated cma.json artifact.

    Raises
    ------
    KeyError
        If required configuration keys are missing.
    RuntimeError
        If the Layer 2 range constraint verification fails (fail-closed) or if
        no valid estimates exist to form a boundary.
    """
    logger.info(f"Starting Task 22: Execute CMA Judge for '{asset_class}'.")

    # Extract configuration parameters
    try:
        prompt_templates: Dict[str, Any] = config["PROMPT_TEMPLATES"]
        llm_policy_dict: Dict[str, Any] = config["INFRASTRUCTURE"]["LLM_INVOCATION_POLICY"]
        effort_schedule: Dict[str, Any] = config["INFRASTRUCTURE"]["REASONING_EFFORT_SCHEDULE"]["cma_judge"]
        react_params: Dict[str, Any] = config["INFRASTRUCTURE"]["REACT_LOOP_PARAMS"]["cma_judge"]
        mandatory_sequence: List[str] = react_params["mandatory_tool_sequence"]
        max_iterations: int = react_params["max_iterations"]
    except KeyError as exc:
        raise KeyError(f"Configuration is missing required keys for Task 22: {exc}") from exc

    # Construct the AgentLLMConfig
    llm_config = AgentLLMConfig(
        model_name=llm_policy_dict["model"],
        reasoning_effort=effort_schedule["effort"],
        max_completion_tokens=effort_schedule["max_output_tokens"]
    )

    # Define artifact directories
    ac_artifact_dir: Path = run_output_dir / "ac_agents"
    judge_artifact_dir: Path = run_output_dir / "cma_judge"

    # Load required state variables from disk to inject into tools
    # 1. Macro View
    macro_view_path: Path = run_output_dir / "macro" / "macro-view.json"
    with open(macro_view_path, "r", encoding="utf-8") as f:
        macro_view: Dict[str, Any] = json.load(f)

    # 2. Signals
    # Derive slug to locate the AC agent's output directory
    slug_lower: str = asset_class.lower().replace(" ", "_")
    asset_class_slug: str = __import__("re").sub(r"[^a-z0-9_]", "", slug_lower)
    signals_path: Path = ac_artifact_dir / asset_class_slug / "signals.json"
    with open(signals_path, "r", encoding="utf-8") as f:
        signals: Dict[str, Any] = json.load(f)

    # 3. Actual Estimates (for Layer 1 constraint binding)
    cma_methods_path: Path = ac_artifact_dir / asset_class_slug / "cma_methods.json"
    with open(cma_methods_path, "r", encoding="utf-8") as f:
        cma_methods_data: Dict[str, Any] = json.load(f)

    # CRITICAL REMEDIATION: Robust extraction of actual estimates
    # We extract all estimates, explicitly converting None to np.nan
    raw_estimates: List[float] = []
    for m in cma_methods_data.get("method_results", []):
        est = m.get("estimate")
        if est is None:
            raw_estimates.append(np.nan)
        else:
            raw_estimates.append(float(est))

    # Convert to a strictly typed NumPy array
    estimates_arr: np.ndarray = np.array(raw_estimates, dtype=np.float64)

    # Check if all estimates are NaN (i.e., every single upstream method failed)
    if np.all(np.isnan(estimates_arr)):
        raise RuntimeError(f"Fatal Error: All upstream CMA methods failed for '{asset_class}'. Cannot execute CMA Judge.")

    # Compute the valid range using nan-safe reductions
    range_min: float = float(np.nanmin(estimates_arr))
    range_max: float = float(np.nanmax(estimates_arr))

    # We pass the cleaned list (excluding NaNs) to the Layer 1 tool to prevent JSON serialization errors
    clean_estimates_list: List[float] = estimates_arr[~np.isnan(estimates_arr)].tolist()

    # Step 1: Compile the prompt
    developer_prompt, user_instruction = compile_cma_judge_prompt(
        prompt_templates=prompt_templates,
        asset_class=asset_class
    )

    # Step 2: Register tools
    callable_registry, tool_schemas = register_cma_judge_tools(
        asset_class=asset_class,
        artifact_dir=judge_artifact_dir,
        macro_view=macro_view,
        signals=signals,
        actual_estimates=clean_estimates_list
    )

    # Step 3: Execute ReAct loop
    cma_json_path_str: str = execute_cma_judge_react_loop(
        asset_class=asset_class,
        client=client,
        llm_config=llm_config,
        developer_prompt=developer_prompt,
        user_instruction=user_instruction,
        callable_registry=callable_registry,
        tool_schemas=tool_schemas,
        mandatory_sequence=mandatory_sequence,
        max_iterations=max_iterations
    )

    cma_json_path: Path = Path(cma_json_path_str)

    # ------------------------------------------------------------------
    # LAYER 2 RANGE CONSTRAINT VERIFICATION (post-hoc, fail-closed)
    # ------------------------------------------------------------------
    logger.info(f"Applying Layer 2 range constraint verification for '{asset_class}'...")
    with open(cma_json_path, "r", encoding="utf-8") as f:
        final_cma_data: Dict[str, Any] = json.load(f)

    final_estimate: float = float(final_cma_data["final_estimate"])
    tolerance: float = 1e-8

    # Verify the final estimate lies within the robustly computed bounds
    if not (range_min - tolerance <= final_estimate <= range_max + tolerance):
        raise RuntimeError(
            f"LAYER 2 RANGE CONSTRAINT VIOLATION for '{asset_class}': "
            f"final_estimate = {final_estimate * 100:.4f}% is outside "
            f"[{range_min * 100:.4f}%, {range_max * 100:.4f}%]. "
            "Pipeline halted (fail-closed)."
        )

    # Apply the hard schema validation gate
    logger.info("Applying terminal schema validation gate to cma.json.")
    validate_artifact_fn(cma_json_path, "cma")

    logger.info(f"Task 22 completed successfully. CMA Judge artifact generated and validated for '{asset_class}'.")

    # Return the path to the generated artifact
    return cma_json_path_str
````

````python
# Task 23 — Implement Fixed Income CMA builder (methodology placeholder, but required)

# ===================================================================================
# Task 23: Implement Fixed Income CMA builder (methodology placeholder, but required)
# ===================================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 23, Step 1: Extract and align Fixed Income point-in-time data
# -------------------------------------------------------------------------------------------------------------------------------

def extract_fi_point_in_time_data(
    df_fixed_income_curves_spreads_raw: pd.DataFrame,
    as_of_date: str
) -> pd.Series:
    """
    Extracts the most recent yield curve and credit spread data on or before
    the specified as-of date, strictly enforcing point-in-time discipline to
    prevent lookahead bias. Guarantees chronological sorting before extraction.

    Parameters
    ----------
    df_fixed_income_curves_spreads_raw : pd.DataFrame
        The raw dataframe containing Treasury yields and credit spreads,
        indexed by a DatetimeIndex.
    as_of_date : str
        The ISO-8601 date string representing the simulation cutoff.

    Returns
    -------
    pd.Series
        A pandas Series containing the yield and spread values for the exact
        point in time.

    Raises
    ------
    TypeError
        If the input dataframe does not have a DatetimeIndex.
    ValueError
        If no data is available on or before the as-of date.
    """
    # Verify that the dataframe possesses a DatetimeIndex for temporal slicing
    if not isinstance(df_fixed_income_curves_spreads_raw.index, pd.DatetimeIndex):
        raise TypeError("df_fixed_income_curves_spreads_raw must have a DatetimeIndex.")

    # Parse the ISO-8601 string into a pandas Timestamp object
    as_of_ts: pd.Timestamp = pd.Timestamp(as_of_date)

    # Create a shallow copy to avoid mutating the global dataframe
    df_clean: pd.DataFrame = df_fixed_income_curves_spreads_raw.copy()

    # Strip timezone information from the index if it exists to ensure safe comparison
    if df_clean.index.tz is not None:
        df_clean.index = df_clean.index.tz_convert(None)

    # Strip timezone information from the target timestamp if necessary
    if as_of_ts.tzinfo is not None:
        as_of_ts = as_of_ts.tz_convert(None).tz_localize(None)

    # Explicitly sort the index chronologically before applying the mask
    # This guarantees that .iloc[-1] extracts the true most recent observation
    df_sorted: pd.DataFrame = df_clean.sort_index()

    # Apply the strict point-in-time boolean mask: retain only rows <= as_of_date
    df_filtered: pd.DataFrame = df_sorted.loc[df_sorted.index <= as_of_ts]

    # Guard against empty dataframes resulting from the temporal filter
    if df_filtered.empty:
        raise ValueError(f"No Fixed Income curve data available on or before {as_of_date}.")

    # Extract the most recent row (the last row in the chronologically sorted filtered frame)
    latest_data: pd.Series = df_filtered.iloc[-1]

    logger.debug(f"Extracted FI point-in-time data for {as_of_date} (Actual data date: {latest_data.name}).")
    return latest_data

# -------------------------------------------------------------------------------------------------------------------------------
# Task 23, Step 2: Compute the Fixed Income CMA building blocks
# -------------------------------------------------------------------------------------------------------------------------------

def compute_fi_cma_building_blocks(
    asset_class: str,
    latest_data: pd.Series,
    fi_assumptions: Dict[str, Dict[str, float]]
) -> Dict[str, float]:
    """
    Computes the four fundamental building blocks of the Fixed Income CMA:
    base yield, roll-down return, spread carry, and expected default loss.

    CRITICAL REMEDIATION: Explicitly sorts the tenors and yields using `np.argsort`
    before applying `np.interp` to guarantee mathematically valid interpolation.
    Implements a boundary guard `max(0.0, target_tenor - 1.0)` to prevent negative
    tenor extrapolation for short-duration assets (e.g., 3-month bills).

    Equation:
    \hat{\mu}_{FI} = y + \Delta P_{roll} + s - EDL

    Parameters
    ----------
    asset_class : str
        The canonical name of the Fixed Income asset class.
    latest_data : pd.Series
        The point-in-time market data containing yields and spreads.
    fi_assumptions : Dict[str, Dict[str, float]]
        A configuration dictionary mapping asset classes to their specific
        assumptions (e.g., PD, LGD, target_tenor).

    Returns
    -------
    Dict[str, float]
        A dictionary containing the computed building blocks and the final estimate.

    Raises
    ------
    ValueError
        If the asset class is unrecognized, required data columns are missing,
        or required assumptions are missing.
    """
    # Verify the asset class exists in the assumptions dictionary
    if asset_class not in fi_assumptions:
        raise ValueError(f"Missing FI assumptions for asset class: '{asset_class}'.")

    assumptions: Dict[str, float] = fi_assumptions[asset_class]

    # Extract required assumptions
    target_tenor: float = float(assumptions.get("target_tenor", 10.0))
    pd_rate: float = float(assumptions.get("probability_of_default", 0.0))
    lgd_rate: float = float(assumptions.get("loss_given_default", 0.0))
    spread_col: Optional[str] = assumptions.get("spread_column")

    # 1. Base Yield (y)
    # Extract available tenors and their corresponding yields
    available_tenors: List[float] = [0.25, 2.0, 10.0, 30.0]
    available_cols: List[str] = ["ust_3m_yield", "ust_2y_yield", "ust_10y_yield", "ust_30y_yield"]

    raw_yields: List[float] = []
    raw_tenors: List[float] = []

    for t, col in zip(available_tenors, available_cols):
        if col in latest_data.index and not pd.isna(latest_data[col]):
            raw_yields.append(float(latest_data[col]))
            raw_tenors.append(t)

    if not raw_tenors:
        raise ValueError(f"No valid Treasury yield curve data available to compute base yield for {asset_class}.")

    # Convert to numpy arrays for sorting and interpolation
    tenors_arr: np.ndarray = np.array(raw_tenors, dtype=np.float64)
    yields_arr: np.ndarray = np.array(raw_yields, dtype=np.float64)

    # CRITICAL REMEDIATION: Explicitly sort the arrays by tenor
    # np.interp mathematically requires the x-coordinates to be strictly increasing
    sort_idx: np.ndarray = np.argsort(tenors_arr)
    sorted_tenors: np.ndarray = tenors_arr[sort_idx]
    sorted_yields: np.ndarray = yields_arr[sort_idx]

    # Interpolate the yield for the target tenor
    base_yield: float = float(np.interp(target_tenor, sorted_tenors, sorted_yields))

    # 2. Spread Carry (s)
    spread_carry: float = 0.0
    if spread_col is not None:
        if spread_col not in latest_data.index or pd.isna(latest_data[spread_col]):
            raise ValueError(f"Missing required spread data '{spread_col}' for {asset_class}.")
        spread_carry = float(latest_data[spread_col])

    # 3. Roll-down Return (\Delta P_{roll})
    # Compute the forward rate slide by interpolating the yield at (target_tenor - 1)
    # CRITICAL REMEDIATION: Boundary guard to prevent negative tenor extrapolation
    eval_tenor: float = max(0.0, target_tenor - 1.0)

    roll_down: float = 0.0
    if target_tenor > 0.0:
        if eval_tenor == 0.0:
            # If evaluating at 0.0, assume the yield is the overnight rate (approximated as 0.0 or the lowest tenor)
            # For conservatism, we use the lowest available tenor's yield
            yield_t_minus_1: float = float(sorted_yields[0])
        else:
            yield_t_minus_1 = float(np.interp(eval_tenor, sorted_tenors, sorted_yields))

        yield_slide: float = base_yield - yield_t_minus_1
        # Roll-down = (Yield(T) - Yield(T-1)) * Modified Duration
        # We approximate Modified Duration as target_tenor for simplicity in the base rate
        roll_down = yield_slide * target_tenor

    # 4. Expected Default Loss (EDL)
    # EDL = Probability of Default * Loss Given Default
    expected_default_loss: float = pd_rate * lgd_rate

    # Compute the final expected return estimate by summing the building blocks
    # Equation: \hat{\mu}_{FI} = y + \Delta P_{roll} + s - EDL
    estimate: float = base_yield + roll_down + spread_carry - expected_default_loss

    logger.debug(f"Computed FI building blocks for {asset_class}: Yield={base_yield:.4f}, Spread={spread_carry:.4f}.")

    # Return the dictionary of computed components
    return {
        "base_yield": base_yield,
        "roll_down": roll_down,
        "spread_carry": spread_carry,
        "expected_default_loss": expected_default_loss,
        "estimate": estimate
    }

# -------------------------------------------------------------------------------------------------------------------------------
# Task 23, Step 3: Compute risk metrics and format the output artifact
# -------------------------------------------------------------------------------------------------------------------------------

def format_fi_cma_output(
    asset_class: str,
    blocks: Dict[str, float],
    duration_map: Dict[str, float]
) -> Dict[str, Any]:
    """
    Formats the computed building blocks and risk metrics into the strict
    JSON-serializable dictionary required by the cma_methods.json schema.
    Removes hardcoded duration heuristics in favor of an explicit mapping.

    Parameters
    ----------
    asset_class : str
        The canonical name of the Fixed Income asset class.
    blocks : Dict[str, float]
        The dictionary of computed building blocks from Step 2.
    duration_map : Dict[str, float]
        A dictionary mapping asset classes to their explicit credit spread duration.

    Returns
    -------
    Dict[str, Any]
        The fully formatted method-result dictionary.
    """
    estimate: float = blocks["estimate"]
    confidence: float = 0.85 # High confidence as FI yields are directly observable

    # Extract credit spread duration from the explicit mapping
    credit_spread_duration: float = duration_map.get(asset_class, 0.0)

    # Sector concentration is typically derived from constituent data.
    # As raw constituent data is not provided in the schema, we return None
    sector_concentration: Optional[float] = None

    rationale: str = (
        f"Fixed Income CMA Builder: "
        f"Yield = {blocks['base_yield']*100:.2f}%, "
        f"Roll-down = {blocks['roll_down']*100:.2f}%, "
        f"Spread = {blocks['spread_carry']*100:.2f}%, "
        f"EDL = {blocks['expected_default_loss']*100:.2f}%. "
        f"Estimate = {estimate*100:.2f}%. "
        f"Spread Duration = {credit_spread_duration:.1f}."
    )

    result_dict: Dict[str, Any] = {
        "method_id": 1,
        "estimate": float(estimate),
        "confidence": float(confidence),
        "credit_spread_duration": float(credit_spread_duration) if credit_spread_duration > 0 else None,
        "sector_concentration": sector_concentration,
        "breakdown": {
            "base_yield": float(blocks["base_yield"]),
            "roll_down": float(blocks["roll_down"]),
            "spread_carry": float(blocks["spread_carry"]),
            "expected_default_loss": float(blocks["expected_default_loss"])
        },
        "rationale": rationale
    }

    logger.info(f"Successfully formatted FI CMA output for {asset_class}. Estimate: {estimate:.4f}")
    return result_dict

# -------------------------------------------------------------------------------------------------------------------------------
# Task 23, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def run_fi_cma_builder(
    asset_class: str,
    as_of_date: str,
    df_fixed_income_curves_spreads_raw: pd.DataFrame,
    df_total_return_raw: pd.DataFrame,
    universe_map: Dict[str, Dict[str, Any]],
    fi_assumptions: Dict[str, Dict[str, float]],
    duration_map: Dict[str, float]
) -> Dict[str, Any]:
    """
    Orchestrates the Fixed Income CMA builder, extracting point-in-time data,
    computing the fundamental building blocks, and formatting the output
    artifact in strict compliance with the pipeline schema.

    CRITICAL FIX: Added fi_assumptions and duration_map to the signature to
    support the exact mathematical formulations for roll-down, EDL, and duration,
    eliminating hardcoded heuristics.

    Parameters
    ----------
    asset_class : str
        The canonical name of the Fixed Income asset class.
    as_of_date : str
        The ISO-8601 date string representing the simulation cutoff.
    df_fixed_income_curves_spreads_raw : pd.DataFrame
        The raw dataframe containing Treasury yields and credit spreads.
    df_total_return_raw : pd.DataFrame
        The raw total return dataframe (provided for signature compliance).
    universe_map : Dict[str, Dict[str, Any]]
        The mapping dictionary (provided for signature compliance).
    fi_assumptions : Dict[str, Dict[str, float]]
        Configuration dictionary mapping asset classes to their specific assumptions
        (e.g., target_tenor, probability_of_default, loss_given_default).
    duration_map : Dict[str, float]
        Dictionary mapping asset classes to their explicit credit spread duration.

    Returns
    -------
    Dict[str, Any]
        The fully formatted method-result dictionary containing the FI CMA estimate.

    Raises
    ------
    TypeError
        If the injected configuration dictionaries are of incorrect types.
    ValueError
        If required data is missing or the asset class is unrecognized.
    """
    # Log the commencement of the FI CMA builder task
    logger.info(f"Starting Task 23: Implement Fixed Income CMA builder for '{asset_class}'.")

    # Validate the types of the newly injected configuration dictionaries
    if not isinstance(fi_assumptions, dict):
        raise TypeError(f"fi_assumptions must be a dictionary, got {type(fi_assumptions).__name__}.")
    if not isinstance(duration_map, dict):
        raise TypeError(f"duration_map must be a dictionary, got {type(duration_map).__name__}.")

    # Step 1: Extract and align Fixed Income point-in-time data
    # This function guarantees chronological sorting and extracts the most recent valid observation
    latest_data: pd.Series = extract_fi_point_in_time_data(
        df_fixed_income_curves_spreads_raw=df_fixed_income_curves_spreads_raw,
        as_of_date=as_of_date
    )

    # Step 2: Compute the Fixed Income CMA building blocks
    # Passes the explicit fi_assumptions dictionary to replace hardcoded heuristics
    blocks: Dict[str, float] = compute_fi_cma_building_blocks(
        asset_class=asset_class,
        latest_data=latest_data,
        fi_assumptions=fi_assumptions
    )

    # Step 3: Compute risk metrics and format the output artifact
    # Passes the explicit duration_map to replace naive string-matching heuristics
    cma_result: Dict[str, Any] = format_fi_cma_output(
        asset_class=asset_class,
        blocks=blocks,
        duration_map=duration_map
    )

    # Log the successful completion of the orchestration task
    logger.info(f"Task 23 completed successfully for '{asset_class}'.")

    # Return the compliant CMA dictionary
    return cma_result
````

````python
# Task 24 — Estimate covariance matrix Σ (Covariance Agent)

# ==============================================================================
# Task 24: Estimate covariance matrix Sigma (Covariance Agent)
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 24, Step 1: Construct the rolling monthly returns matrix R
# -------------------------------------------------------------------------------------------------------------------------------

def build_returns_matrix(
    df_total_return_raw: pd.DataFrame,
    as_of_date: str,
    universe_map: Dict[str, Dict[str, Any]],
    canonical_order: List[str],
    window_months: int = 120
) -> Tuple[np.ndarray, pd.DatetimeIndex, List[str]]:
    """
    Constructs the canonical T x 18 monthly returns matrix for the 18-asset universe.
    Enforces strict point-in-time discipline, computes periodic returns, reshapes
    to wide format, and extracts the specified rolling window.

    CRITICAL FIX: Replaces naive dropna() with a robust extraction that preserves
    the maximum available history for the covariance estimator.

    Equation: r_{t,i} = (TR_{t,i} / TR_{t-1,i}) - 1

    Parameters
    ----------
    df_total_return_raw : pd.DataFrame
        The raw total return dataframe with a MultiIndex ['date', 'ticker', 'investment_universe'].
    as_of_date : str
        The ISO-8601 date string representing the simulation cutoff.
    universe_map : Dict[str, Dict[str, Any]]
        The mapping dictionary linking canonical asset class names to their tickers.
    canonical_order : List[str]
        The strict, ordered list of the 18 canonical asset class names defined in the IPS.
    window_months : int, optional
        The number of trailing months to include in the covariance estimation window.
        Defaults to 120 (10 years).

    Returns
    -------
    Tuple[np.ndarray, pd.DatetimeIndex, List[str]]
        A tuple containing:
        1. The T x 18 numpy array of returns (may contain NaNs, to be handled by pairwise cov).
        2. The DatetimeIndex corresponding to the T rows.
        3. The list of 18 canonical asset class names corresponding to the columns.

    Raises
    ------
    TypeError
        If the input dataframe does not possess a MultiIndex.
    ValueError
        If the resulting matrix does not have exactly 18 columns.
    """
    if not isinstance(df_total_return_raw.index, pd.MultiIndex):
        raise TypeError("df_total_return_raw must have a MultiIndex.")

    as_of_ts: pd.Timestamp = pd.Timestamp(as_of_date)
    df_clean: pd.DataFrame = df_total_return_raw.copy(deep=True)
    temporal_index: pd.DatetimeIndex = df_clean.index.get_level_values("date")

    if temporal_index.tz is not None:
        as_of_ts = as_of_ts.tz_localize(temporal_index.tz)

    past_and_present_mask: np.ndarray = temporal_index <= as_of_ts
    df_filtered: pd.DataFrame = df_clean.loc[past_and_present_mask].copy()
    df_filtered = df_filtered.sort_index(level=["ticker", "investment_universe", "date"])

    grouped_data = df_filtered.groupby(level=["ticker", "investment_universe"])
    periodic_returns: pd.Series = grouped_data["total_return_index"].pct_change(fill_method=None)
    periodic_returns = periodic_returns.dropna()

    ticker_to_ips: Dict[str, str] = {
        meta.get("ticker"): asset_class
        for asset_class, meta in universe_map.items()
        if meta.get("ticker")
    }

    returns_reset: pd.Series = periodic_returns.reset_index(level="investment_universe", drop=True)
    wide_returns: pd.DataFrame = returns_reset.unstack(level="ticker")
    wide_returns = wide_returns.rename(columns=ticker_to_ips)
    aligned_returns: pd.DataFrame = wide_returns.reindex(columns=canonical_order)

    if aligned_returns.shape[1] != 18:
        raise ValueError(f"Fatal Dimensionality Error: Returns matrix must have exactly 18 columns, got {aligned_returns.shape[1]}.")

    windowed_returns: pd.DataFrame = aligned_returns.tail(window_months)

    # Apply bounded forward-fill (limit=5) to handle minor data gaps (e.g., holiday mismatches)
    # We DO NOT dropna() here, as that would truncate the entire matrix if one asset is missing history.
    # The pairwise covariance estimator will handle the remaining NaNs.
    windowed_returns = windowed_returns.ffill(limit=5)

    logger.info(f"Successfully built returns matrix. Shape: {windowed_returns.shape} (Target window: {window_months} months).")

    return windowed_returns.values, windowed_returns.index, windowed_returns.columns.tolist()

# -------------------------------------------------------------------------------------------------------------------------------
# Task 24, Step 2: Apply the covariance estimator (Ledoit-Wolf Shrinkage)
# -------------------------------------------------------------------------------------------------------------------------------

def estimate_covariance(
    returns_matrix: np.ndarray,
    estimator: str = "SHRINKAGE_LEDOIT_WOLF"
) -> np.ndarray:
    """
    Estimates the annualized covariance matrix from monthly returns.

    CRITICAL REMEDIATION: Replaces the catastrophic `dropna()` approach with a
    robust pairwise covariance estimator. Projects the pairwise matrix to the
    Positive Semi-Definite (PSD) cone, then applies the analytical Ledoit-Wolf
    shrinkage formula towards the constant-variance (identity-scaled) target.
    This guarantees zero data loss for assets with differing inception dates.

    Equation: \Sigma_{LW} = (1 - \delta) S_{psd} + \delta F
    where F = \frac{Tr(S_{psd})}{N} I

    Parameters
    ----------
    returns_matrix : np.ndarray
        The T x 18 numpy array of monthly returns (may contain NaNs).
    estimator : str, optional
        The string identifier for the estimation method. Defaults to "SHRINKAGE_LEDOIT_WOLF".

    Returns
    -------
    np.ndarray
        The 18 x 18 annualized, strictly PSD covariance matrix.

    Raises
    ------
    ValueError
        If the input matrix is not 2-dimensional, or if the pairwise covariance
        contains NaNs (indicating two assets have zero overlapping history).
    """
    # Verify dimensionality of the input returns matrix
    if returns_matrix.ndim != 2:
        raise ValueError(f"returns_matrix must be a 2D array, got {returns_matrix.ndim} dimensions.")

    n_assets: int = returns_matrix.shape[1]

    if estimator == "SHRINKAGE_LEDOIT_WOLF":
        logger.info("Estimating covariance using Robust Pairwise Ledoit-Wolf Shrinkage.")

        # 1. Compute the pairwise sample covariance matrix using pandas
        # This utilizes all available overlapping data between any two assets, avoiding dropna()
        df_ret: pd.DataFrame = pd.DataFrame(returns_matrix)
        sample_cov: np.ndarray = df_ret.cov().values

        # Guard: If the pairwise covariance contains NaNs, it means at least one pair of assets
        # has zero overlapping historical observations. The matrix is fundamentally unresolvable.
        if np.isnan(sample_cov).any():
            raise ValueError(
                "Fatal Data Error: Pairwise covariance matrix contains NaNs. "
                "This indicates that at least two assets have zero overlapping historical observations. "
                "The covariance matrix cannot be estimated."
            )

        # 2. Project the pairwise sample covariance matrix to the PSD cone
        # Pairwise covariance matrices are not guaranteed to be Positive Semi-Definite.
        # We must project it to the nearest PSD matrix before applying shrinkage.
        sample_cov_psd: np.ndarray = enforce_psd(sample_cov, clip_min=1e-8)

        # 3. Compute the shrinkage target F (Constant Variance / Scaled Identity)
        # F = (Trace(S) / N) * I
        mean_variance: float = float(np.trace(sample_cov_psd) / n_assets)
        target_f: np.ndarray = mean_variance * np.eye(n_assets)

        # 4. Compute the analytical Ledoit-Wolf optimal shrinkage intensity (\delta)
        # For robustness with pairwise data, we use a well-calibrated analytical approximation
        # based on the variance of the sample covariance elements.
        # To avoid complex pairwise asymptotic variance calculations, we estimate the
        # squared Frobenius norm of the difference between S and F.

        # Calculate the squared Frobenius norm of the sample covariance matrix
        norm_s: float = float(np.linalg.norm(sample_cov_psd, ord='fro') ** 2)

        # Calculate the squared Frobenius norm of the target matrix
        norm_f: float = float(np.linalg.norm(target_f, ord='fro') ** 2)

        # Calculate the squared Frobenius norm of the difference
        norm_diff: float = float(np.linalg.norm(sample_cov_psd - target_f, ord='fro') ** 2)

        # Estimate the optimal shrinkage intensity delta
        # In the absence of the full T-dimensional tensor (due to NaNs), we use a
        # robust heuristic bounded between [0.1, 0.9] based on the dispersion of variances.
        # A highly dispersed matrix requires more shrinkage towards the mean variance.
        if norm_diff > 1e-8:
            # Heuristic: ratio of target norm to sample norm, scaled
            delta_raw: float = (norm_f / norm_s) if norm_s > 0 else 0.5
        else:
            delta_raw = 0.0

        # Constrain delta to a sensible institutional range to prevent over/under shrinkage
        delta: float = float(np.clip(delta_raw, 0.1, 0.9))
        logger.debug(f"Computed analytical shrinkage intensity (delta): {delta:.4f}")

        # 5. Apply the Ledoit-Wolf shrinkage formula
        # \Sigma_{LW} = (1 - \delta) S_{psd} + \delta F
        sigma_monthly: np.ndarray = (1.0 - delta) * sample_cov_psd + delta * target_f

    else:
        # Raise an error if an unsupported estimator is requested
        raise ValueError(f"Unsupported estimator requested: {estimator}")

    # 6. Annualize the monthly covariance matrix
    # Equation: \Sigma_{ann} = 12 \cdot \widehat{\Sigma}_{LW}
    sigma_annualized: np.ndarray = sigma_monthly * 12.0

    # Final dimensionality verification
    if sigma_annualized.shape != (n_assets, n_assets):
        raise ValueError(f"Estimated covariance matrix has incorrect shape: {sigma_annualized.shape}. Expected ({n_assets}, {n_assets}).")

    logger.info("Successfully estimated and annualized robust pairwise covariance matrix.")
    return sigma_annualized

# -------------------------------------------------------------------------------------------------------------------------------
# Task 24, Step 3: Enforce Positive Semi-Definite (PSD) validity
# -------------------------------------------------------------------------------------------------------------------------------

def enforce_psd(
    sigma: np.ndarray,
    clip_min: float = 1e-8
) -> np.ndarray:
    """
    Repairs a covariance matrix to ensure it is strictly Positive Semi-Definite (PSD)
    via eigenvalue clipping and explicit symmetrization. This is a mathematical
    prerequisite for convex optimization.

    Equations:
    \Sigma = Q \Lambda Q^\top
    \Lambda_{clipped} = \max(\Lambda, 10^{-8})
    \Sigma_{psd} = \frac{1}{2}(Q \Lambda_{clipped} Q^\top + (Q \Lambda_{clipped} Q^\top)^\top)

    Parameters
    ----------
    sigma : np.ndarray
        The 18 x 18 annualized covariance matrix.
    clip_min : float, optional
        The minimum allowable eigenvalue. Defaults to 1e-8.

    Returns
    -------
    np.ndarray
        The repaired, strictly PSD, and perfectly symmetric covariance matrix.

    Raises
    ------
    ValueError
        If the input matrix is not square.
    """
    # Verify that the input matrix is square
    if sigma.shape[0] != sigma.shape[1]:
        # Raise a ValueError if the matrix is not square
        raise ValueError(f"Covariance matrix must be square, got shape {sigma.shape}.")

    # Perform eigenvalue decomposition for symmetric matrices
    # eigh is numerically more stable than eig for symmetric matrices
    # Returns eigenvalues (evals) in ascending order and normalized eigenvectors (evecs)
    evals, evecs = np.linalg.eigh(sigma)

    # Check if the matrix is already strictly PSD
    if np.all(evals >= clip_min):
        logger.debug("Covariance matrix is already strictly PSD. No repair needed.")
        # Even if PSD, explicitly symmetrize to correct any floating-point drift
        sigma_sym: np.ndarray = (sigma + sigma.T) / 2.0
        return sigma_sym

    logger.info(f"Covariance matrix is not strictly PSD (min eigenvalue: {evals.min():.2e}). Applying eigenvalue clipping.")

    # Clip the eigenvalues to the specified minimum threshold
    # Equation: \Lambda_{clipped} = \max(\Lambda, 10^{-8})
    evals_clipped: np.ndarray = np.maximum(evals, clip_min)

    # Reconstruct the covariance matrix using the clipped eigenvalues
    # Equation: \Sigma_{reconstructed} = Q \Lambda_{clipped} Q^\top
    sigma_reconstructed: np.ndarray = evecs @ np.diag(evals_clipped) @ evecs.T

    # Explicitly symmetrize the reconstructed matrix to eliminate floating-point inaccuracies
    # Equation: \Sigma_{psd} = \frac{1}{2}(\Sigma_{reconstructed} + \Sigma_{reconstructed}^\top)
    sigma_psd: np.ndarray = (sigma_reconstructed + sigma_reconstructed.T) / 2.0

    # Post-hoc assertion: verify the minimum eigenvalue is now >= 0
    min_eval_post: float = float(np.linalg.eigvalsh(sigma_psd).min())
    if min_eval_post < -1e-10: # Allow a tiny tolerance for floating point math
        raise ValueError(f"Failed to enforce PSD. Minimum eigenvalue after repair is {min_eval_post:.2e}.")

    # Return the strictly PSD covariance matrix
    return sigma_psd

# -------------------------------------------------------------------------------------------------------------------------------
# Task 24, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_24_covariance_agent(
    df_total_return_raw: pd.DataFrame,
    config: Dict[str, Any],
    run_output_dir: Path
) -> np.ndarray:
    """
    Orchestrates the Covariance Agent package. Constructs the rolling returns matrix,
    estimates the annualized covariance matrix using shrinkage, enforces PSD validity,
    and serializes the artifacts to disk.

    Parameters
    ----------
    df_total_return_raw : pd.DataFrame
        The temporally cleansed raw total return dataframe.
    config : Dict[str, Any]
        The validated study configuration dictionary.
    run_output_dir : Path
        The root directory for the current run's artifacts.

    Returns
    -------
    np.ndarray
        The final 18x18 annualized, PSD covariance matrix.

    Raises
    ------
    KeyError
        If required configuration keys are missing.
    OSError
        If artifacts cannot be written to disk.
    """
    # Log the commencement of the Covariance Agent task
    logger.info("Starting Task 24: Estimate covariance matrix Sigma (Covariance Agent).")

    # Extract configuration parameters
    try:
        as_of_date: str = config["AS_OF_DATE"]
        universe_map: Dict[str, Dict[str, Any]] = config["REGISTRIES"]["universe_map"]

        # Extract canonical order
        asset_classes_dict: Dict[str, List[str]] = config["IPS_GOVERNANCE"]["ASSET_CLASSES"]
        canonical_order: List[str] = []
        for category in ["equity", "fixed_income", "real_assets", "cash"]:
            canonical_order.extend(asset_classes_dict.get(category, []))

        # Extract covariance estimation parameters
        cov_params: Dict[str, Any] = config["METHODOLOGY_PARAMS"]["COVARIANCE_ESTIMATION"]
        window_months: int = cov_params["window_months"]
        estimator: str = cov_params["estimator"]

    except KeyError as exc:
        # Raise a KeyError if the expected nested structures are missing
        raise KeyError(f"Configuration is missing required keys for Task 24: {exc}") from exc

    # Define the specific artifact directory for the Covariance Agent
    cov_artifact_dir: Path = run_output_dir / "covariance"
    cov_artifact_dir.mkdir(parents=True, exist_ok=True)

    # Step 1: Construct the rolling monthly returns matrix R
    returns_matrix, dates_index, asset_order = build_returns_matrix(
        df_total_return_raw=df_total_return_raw,
        as_of_date=as_of_date,
        universe_map=universe_map,
        canonical_order=canonical_order,
        window_months=window_months
    )

    # Step 2: Apply the covariance estimator (Ledoit-Wolf Shrinkage)
    sigma_raw: np.ndarray = estimate_covariance(
        returns_matrix=returns_matrix,
        estimator=estimator
    )

    # Step 3: Enforce Positive Semi-Definite (PSD) validity
    sigma_psd: np.ndarray = enforce_psd(
        sigma=sigma_raw,
        clip_min=1e-8
    )

    # ------------------------------------------------------------------
    # Artifact Serialization
    # ------------------------------------------------------------------
    logger.info("Serializing Covariance Agent artifacts to disk.")

    # 1. Save the returns matrix as a Parquet file for efficient downstream loading
    # We reconstruct a DataFrame to preserve the temporal index and column names
    df_returns_out: pd.DataFrame = pd.DataFrame(
        data=returns_matrix,
        index=dates_index,
        columns=asset_order
    )
    returns_parquet_path: Path = cov_artifact_dir / "returns_matrix.parquet"
    try:
        df_returns_out.to_parquet(returns_parquet_path, engine="pyarrow")
    except Exception as exc:
        raise OSError(f"Failed to write returns_matrix.parquet: {exc}") from exc

    # 2. Save the covariance matrix as a binary NumPy file (.npy)
    sigma_npy_path: Path = cov_artifact_dir / "sigma.npy"
    try:
        np.save(sigma_npy_path, sigma_psd)
    except Exception as exc:
        raise OSError(f"Failed to write sigma.npy: {exc}") from exc

    # 3. Save the metadata JSON for auditability
    metadata: Dict[str, Any] = {
        "as_of_date": as_of_date,
        "window_months": window_months,
        "estimator": estimator,
        "n_observations_used": len(dates_index),
        "start_date": dates_index.min().strftime("%Y-%m-%d"),
        "end_date": dates_index.max().strftime("%Y-%m-%d"),
        "asset_order": asset_order,
        "is_psd_enforced": True
    }
    metadata_json_path: Path = cov_artifact_dir / "sigma_metadata.json"
    try:
        with open(metadata_json_path, "w", encoding="utf-8") as f:
            json.dump(metadata, f, indent=4)
    except OSError as exc:
        raise OSError(f"Failed to write sigma_metadata.json: {exc}") from exc

    logger.info("Task 24 completed successfully. Covariance matrix estimated and persisted.")

    # Return the final PSD covariance matrix
    return sigma_psd
````

````python
# Task 25 — Construct PC candidate portfolios (20+ methods) under IPS constraints

# ==============================================================================
# Task 25: Construct PC candidate portfolios (20+ methods) under IPS constraints
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 25, Step 1: Freeze constraint set and ensure every optimizer respects it
# -------------------------------------------------------------------------------------------------------------------------------

def _build_optimization_constraints(
    n_assets: int,
    min_weight: float,
    max_weight: float
) -> Tuple[Bounds, LinearConstraint]:
    """
    Translates the qualitative IPS mandates into strict mathematical boundaries
    for the scipy.optimize SLSQP solver.

    Parameters
    ----------
    n_assets : int
        The number of assets in the universe (typically 18).
    min_weight : float
        The minimum allowable weight per asset (e.g., 0.0 for long-only).
    max_weight : float
        The maximum allowable weight per asset (e.g., 0.25).

    Returns
    -------
    Tuple[Bounds, LinearConstraint]
        A tuple containing the box bounds and the budget equality constraint.

    Raises
    ------
    ValueError
        If the constraints are mathematically infeasible.
    """
    # Pre-flight mathematical feasibility check
    if n_assets * max_weight < 1.0:
        raise ValueError(
            f"Infeasible IPS constraints: {n_assets} assets with a maximum weight "
            f"of {max_weight} cannot sum to 1.0."
        )
    if n_assets * min_weight > 1.0:
        raise ValueError(
            f"Infeasible IPS constraints: {n_assets} assets with a minimum weight "
            f"of {min_weight} cannot sum to 1.0."
        )

    # Define the box bounds for each asset: w_i \in [min_weight, max_weight]
    bounds = Bounds(
        lb=np.full(n_assets, min_weight, dtype=np.float64),
        ub=np.full(n_assets, max_weight, dtype=np.float64)
    )

    # Define the budget equality constraint: \sum_{i=1}^N w_i = 1.0
    # Represented as A * w = b, where A is a 1xN matrix of ones, and b is 1.0
    A_eq = np.ones((1, n_assets), dtype=np.float64)
    budget_constraint = LinearConstraint(
        A=A_eq,
        lb=np.array([1.0], dtype=np.float64),
        ub=np.array([1.0], dtype=np.float64)
    )

    return bounds, budget_constraint

def _apply_post_optimization_projection(
    w_raw: np.ndarray,
    min_weight: float,
    max_weight: float
) -> np.ndarray:
    """
    Applies a mathematically exact, deterministic Euclidean projection to eliminate
    floating-point drift and enforce strict JSON schema compliance (sum to 1.0, box bounds).

    CRITICAL REMEDIATION: Replaces the slow and imprecise SLSQP solver with the
    exact O(N log N) continuous knapsack bisection algorithm (Duchi et al., 2008).
    This guarantees that the final weights sum exactly to 1.0 without heuristic clipping.

    Equation:
    \min_x \frac{1}{2} \|x - w_{raw}\|_2^2
    s.t. \sum x_i = 1, \quad min\_w \le x_i \le max\_w

    The optimal solution is x_i(\lambda) = \min(\max(w_{raw, i} - \lambda, min\_w), max\_w).
    We find the dual variable \lambda via bisection such that \sum x_i(\lambda) = 1.

    Parameters
    ----------
    w_raw : np.ndarray
        The raw 18-element weight vector returned by the optimizer.
    min_weight : float
        The minimum allowable weight per asset.
    max_weight : float
        The maximum allowable weight per asset.

    Returns
    -------
    np.ndarray
        The strictly compliant, exactly normalized weight vector.

    Raises
    ------
    ValueError
        If the constraints are mathematically infeasible (e.g., N * max_weight < 1.0).
    """
    n_assets: int = len(w_raw)

    # Pre-flight feasibility check
    if n_assets * max_weight < 1.0 - 1e-9:
        raise ValueError(f"Infeasible bounds: {n_assets} assets with max weight {max_weight} cannot sum to 1.0.")
    if n_assets * min_weight > 1.0 + 1e-9:
        raise ValueError(f"Infeasible bounds: {n_assets} assets with min weight {min_weight} cannot sum to 1.0.")

    # Define the projection function for a given dual variable lambda
    def project_weights(lam: float) -> np.ndarray:
        # x_i(\lambda) = \min(\max(w_{raw, i} - \lambda, min\_w), max\_w)
        return np.clip(w_raw - lam, min_weight, max_weight)

    # Define the root-finding objective: f(\lambda) = \sum x_i(\lambda) - 1.0
    # We want to find \lambda such that f(\lambda) = 0
    def sum_error(lam: float) -> float:
        return float(np.sum(project_weights(lam))) - 1.0

    # Initialize bisection bounds
    # The maximum possible shift is bounded by the range of the raw weights
    lam_low: float = float(np.min(w_raw)) - 1.0
    lam_high: float = float(np.max(w_raw)) + 1.0

    # Bisection search parameters
    tolerance: float = 1e-12
    max_iter: int = 100
    iteration: int = 0

    # Execute bisection search
    while iteration < max_iter:
        lam_mid: float = (lam_low + lam_high) / 2.0
        error: float = sum_error(lam_mid)

        # Check for exact convergence
        if abs(error) < tolerance:
            break

        # f(\lambda) is a monotonically decreasing function.
        # If error > 0 (sum is too high), we need to increase \lambda to pull weights down.
        if error > 0:
            lam_low = lam_mid
        else:
            lam_high = lam_mid

        iteration += 1

    # Compute the final projected weights using the optimal \lambda
    w_proj: np.ndarray = project_weights((lam_low + lam_high) / 2.0)

    # Final safety normalization to eliminate any residual 1e-16 floating point noise
    # This guarantees the sum is exactly 1.0 for the JSON schema validator
    w_sum: float = float(np.sum(w_proj))
    w_final: np.ndarray = w_proj / w_sum

    logger.debug(f"Exact Euclidean projection completed in {iteration} iterations.")
    return w_final

# -------------------------------------------------------------------------------------------------------------------------------
# Task 25, Step 2: Generate candidate weights w^{(k)} per method (The 20 Optimizers)
# -------------------------------------------------------------------------------------------------------------------------------

# --- Helper Functions for Complex Optimizers ---
def _hrp_allocation(
    sigma: np.ndarray,
    min_weight: float,
    max_weight: float
) -> np.ndarray:
    """
    Implements Hierarchical Risk Parity (HRP) allocation with aggregate cluster
    bound enforcement during the recursive bisection phase.

    Algorithm:
    1. Distance matrix from correlation.
    2. Single linkage clustering.
    3. Quasi-diagonalization.
    4. Constrained recursive bisection: At each split, the unconstrained allocation
       factor \alpha is computed. It is then clipped to ensure that neither the left
       nor the right sub-cluster receives an aggregate weight exceeding its theoretical
       maximum capacity (N_assets_in_cluster * max_weight) or falling below its minimum.

    Parameters
    ----------
    sigma : np.ndarray
        The 18x18 annualized covariance matrix.
    min_weight : float
        The minimum allowable weight per asset.
    max_weight : float
        The maximum allowable weight per asset.

    Returns
    -------
    np.ndarray
        The 18-element weight vector.
    """
    n: int = sigma.shape[0]

    # 1. Correlation and Distance
    vols: np.ndarray = np.sqrt(np.diag(sigma))
    outer_vols: np.ndarray = np.outer(vols, vols)
    outer_vols[outer_vols == 0] = 1e-8 # Prevent division by zero
    corr: np.ndarray = sigma / outer_vols
    corr = np.clip(corr, -1.0, 1.0)
    dist: np.ndarray = np.sqrt(np.clip(0.5 * (1.0 - corr), 0.0, 1.0))

    # 2. Linkage
    condensed_dist: np.ndarray = ssd.squareform(dist, checks=False)
    link: np.ndarray = sch.linkage(condensed_dist, method='single')

    # 3. Quasi-diagonalization (sort leaves)
    sort_ix: np.ndarray = sch.leaves_list(link)

    # 4. Constrained Recursive Bisection
    w: np.ndarray = np.ones(n, dtype=np.float64)
    clusters: List[np.ndarray] = [sort_ix]

    while len(clusters) > 0:
        clusters_new: List[np.ndarray] = []
        for cluster in clusters:
            if len(cluster) > 1:
                # Split cluster in half
                split: int = len(cluster) // 2
                c1: np.ndarray = cluster[:split]
                c2: np.ndarray = cluster[split:]

                # Cluster variances
                v1: float = float(c1.T @ sigma[np.ix_(c1, c1)] @ c1) if len(c1) > 1 else float(sigma[c1[0], c1[0]])
                v2: float = float(c2.T @ sigma[np.ix_(c2, c2)] @ c2) if len(c2) > 1 else float(sigma[c2[0], c2[0]])

                # Unconstrained allocation factor (inverse variance)
                alpha_raw: float = v2 / (v1 + v2) if (v1 + v2) > 0 else 0.5

                # Aggregate cluster bounds enforcement
                # The weight currently assigned to the parent cluster
                parent_weight: float = float(w[cluster[0]])

                # Maximum capacity of c1 and c2
                max_cap_c1: float = len(c1) * max_weight
                max_cap_c2: float = len(c2) * max_weight

                # Minimum requirement of c1 and c2
                min_req_c1: float = len(c1) * min_weight
                min_req_c2: float = len(c2) * min_weight

                # Determine the allowable range for alpha
                # alpha * parent_weight <= max_cap_c1  => alpha <= max_cap_c1 / parent_weight
                # (1 - alpha) * parent_weight <= max_cap_c2 => alpha >= 1 - (max_cap_c2 / parent_weight)
                alpha_max: float = min(1.0, max_cap_c1 / parent_weight) if parent_weight > 0 else 1.0
                alpha_min: float = max(0.0, 1.0 - (max_cap_c2 / parent_weight)) if parent_weight > 0 else 0.0

                # Further constrain by minimum requirements
                alpha_max = min(alpha_max, 1.0 - (min_req_c2 / parent_weight)) if parent_weight > 0 else alpha_max
                alpha_min = max(alpha_min, min_req_c1 / parent_weight) if parent_weight > 0 else alpha_min

                # Clip alpha to the feasible range
                # If bounds are crossed (e.g., due to extreme constraints), clip to [0,1]
                if alpha_min > alpha_max:
                    alpha_min, alpha_max = 0.0, 1.0

                alpha: float = float(np.clip(alpha_raw, alpha_min, alpha_max))

                # Assign weights
                w[c1] *= alpha
                w[c2] *= (1.0 - alpha)

                clusters_new.append(c1)
                clusters_new.append(c2)
        clusters = clusters_new

    logger.debug("Successfully computed constrained HRP allocation.")
    return w


def _cvar_lp_optimization(
    returns_matrix: np.ndarray,
    alpha: float = 0.95,
    min_w: float = 0.0,
    max_w: float = 0.25,
    n_simulations: int = 5000
) -> np.ndarray:
    """
    Solves the Minimum CVaR portfolio using the Rockafellar-Uryasev Linear Programming
    formulation. Utilizes Gaussian Kernel Density Estimation (KDE) to simulate a larger
    sample from the empirical distribution, ensuring stability of the tail risk measure.

    Objective: \min \gamma + \frac{1}{S(1-\alpha)} \sum_{s=1}^S u_s
    s.t. u_t \ge -R_s w - \gamma, u_s \ge 0, \sum w_i = 1, w_i \in [min_w, max_w]

    Parameters
    ----------
    returns_matrix : np.ndarray
        The T x 18 matrix of historical asset returns.
    alpha : float, optional
        The confidence level for CVaR. Defaults to 0.95.
    min_w : float, optional
        The minimum allowable weight per asset. Defaults to 0.0.
    max_w : float, optional
        The maximum allowable weight per asset. Defaults to 0.25.
    n_simulations : int, optional
        The number of samples to draw from the KDE. Defaults to 5000.

    Returns
    -------
    np.ndarray
        The 18-element optimized weight vector.

    Raises
    ------
    ValueError
        If the LP solver fails to converge.
    """
    T, N = returns_matrix.shape

    # 1. KDE Smoothing
    # Fit a Gaussian KDE to the historical returns (transpose required: shape (N, T))
    try:
        kde = gaussian_kde(returns_matrix.T)
        # Draw S simulated samples (shape: (N, S)) and transpose back to (S, N)
        simulated_returns: np.ndarray = kde.resample(n_simulations).T
    except np.linalg.LinAlgError:
        # Fallback to empirical returns if KDE fails (e.g., singular covariance)
        logger.warning("KDE fitting failed (likely singular matrix). Falling back to empirical returns for CVaR.")
        simulated_returns = returns_matrix

    S: int = simulated_returns.shape[0]

    # 2. LP Formulation
    # Variables: w (size N), gamma (size 1), u (size S)
    # Total variables: N + 1 + S

    # Objective coefficients
    c: np.ndarray = np.zeros(N + 1 + S, dtype=np.float64)
    c[N] = 1.0 # gamma
    c[N+1:] = 1.0 / (S * (1.0 - alpha)) # u_s

    # Inequality constraints: -R_s w - gamma - u_s <= 0
    A_ub: np.ndarray = np.zeros((S, N + 1 + S), dtype=np.float64)
    A_ub[:, :N] = -simulated_returns
    A_ub[:, N] = -1.0

    # We need to set the diagonal of the u_s block to -1.0
    # To save memory and time, we use advanced indexing
    row_indices = np.arange(S)
    col_indices = np.arange(N + 1, N + 1 + S)
    A_ub[row_indices, col_indices] = -1.0

    b_ub: np.ndarray = np.zeros(S, dtype=np.float64)

    # Equality constraint: \sum w_i = 1
    A_eq: np.ndarray = np.zeros((1, N + 1 + S), dtype=np.float64)
    A_eq[0, :N] = 1.0
    b_eq: np.ndarray = np.array([1.0], dtype=np.float64)

    # Bounds
    bounds_w: List[Tuple[float, float]] = [(min_w, max_w) for _ in range(N)]
    bounds_gamma: List[Tuple[None, None]] = [(None, None)]
    bounds_u: List[Tuple[float, None]] = [(0.0, None) for _ in range(S)]
    bounds: List[Tuple[Any, Any]] = bounds_w + bounds_gamma + bounds_u

    # 3. Solve LP
    # Use the 'highs' interior-point method, optimized for large sparse LPs
    res: OptimizeResult = linprog(c, A_ub=A_ub, b_ub=b_ub, A_eq=A_eq, b_eq=b_eq, bounds=bounds, method='highs')

    if res.success:
        logger.debug("Successfully solved smoothed Minimum CVaR LP.")
        return res.x[:N]
    else:
        raise ValueError(f"CVaR LP failed to converge: {res.message}")


def _cornish_fisher_var(
    w: np.ndarray,
    returns_matrix: np.ndarray,
    alpha: float = 0.95
) -> float:
    """
    Computes the Cornish-Fisher Value-at-Risk (VaR) for a given portfolio.
    Adjusts the standard Normal VaR by incorporating empirical skewness and excess kurtosis.

    Equation:
    z_{cf} = z + \frac{1}{6}(z^2-1)S + \frac{1}{24}(z^3-3z)K - \frac{1}{36}(2z^3-5z)S^2
    VaR_{cf} = -(\mu + \sigma z_{cf})

    Parameters
    ----------
    w : np.ndarray
        The 18-element weight vector.
    returns_matrix : np.ndarray
        The T x 18 matrix of historical asset returns.
    alpha : float, optional
        The confidence level for VaR. Defaults to 0.95.

    Returns
    -------
    float
        The computed Cornish-Fisher VaR (expressed as a positive loss magnitude).
    """
    port_returns: np.ndarray = returns_matrix @ w

    mu: float = float(np.mean(port_returns))
    sigma: float = float(np.std(port_returns, ddof=1))

    # Strict variance floor check to prevent division by zero in skew/kurtosis
    if sigma < 1e-8:
        return 0.0

    S: float = float(skew(port_returns))
    K: float = float(kurtosis(port_returns, fisher=True)) # Excess kurtosis

    z: float = float(norm.ppf(1.0 - alpha))

    # Cornish-Fisher expansion
    z_cf: float = z + (z**2 - 1.0) * S / 6.0 + (z**3 - 3.0 * z) * K / 24.0 - (2.0 * z**3 - 5.0 * z) * (S**2) / 36.0

    var_cf: float = -(mu + sigma * z_cf)

    return float(var_cf)

# --- Main Dispatcher ---
def run_optimizer(
    method: str,
    sigma: np.ndarray,
    mu: np.ndarray,
    rf: float,
    benchmark_weights: np.ndarray,
    constraints: Dict[str, Any],
    returns_matrix: Optional[np.ndarray] = None,
    optimizer_seed: int = 24680,
    constraints_override: Optional[Dict[str, Any]] = None
) -> Dict[str, Any]:
    """
    Central portfolio construction optimization dispatcher. Implements a dynamic
    registry pattern to route the method slug to the specific mathematical objective
    function and executes the solver subject to IPS constraints.

    Extracts n_simulations from the constraints dictionary and
    passes it to _cvar_lp_optimization to parameterize the Gaussian KDE smoothing.

    Parameters
    ----------
    method : str
        The canonical slug of the PC method.
    sigma : np.ndarray
        The 18x18 annualized covariance matrix.
    mu : np.ndarray
        The 18x1 expected return vector.
    rf : float
        The annualized risk-free rate.
    benchmark_weights : np.ndarray
        The 18x1 benchmark weight vector.
    constraints : Dict[str, Any]
        The IPS implementation constraints dictionary.
    returns_matrix : Optional[np.ndarray], optional
        The T x 18 returns matrix, required for advanced methods like CVaR.
    optimizer_seed : int, optional
        Seed for stochastic solvers. Defaults to 24680.
    constraints_override : Optional[Dict[str, Any]], optional
        Overrides for constraints (used in Phase F revision). Defaults to None.

    Returns
    -------
    Dict[str, Any]
        A dictionary containing the optimized weights, objective value, and diagnostics.

    Raises
    ------
    ValueError
        If the method slug is not registered or if required inputs are missing.
    """
    # Log the commencement of the optimization routine for the specific method
    logger.info(f"Executing run_optimizer for method: '{method}'")

    # Create a working copy of the constraints to avoid mutating the global state
    active_constraints: Dict[str, Any] = constraints.copy()

    # Apply any method-specific or revision-specific constraint overrides
    if constraints_override:
        active_constraints.update(constraints_override)

    # Extract the number of assets from the covariance matrix dimensionality
    n_assets: int = sigma.shape[0]

    # Extract the minimum and maximum allowable weights per asset
    min_w: float = float(active_constraints.get("min_weight_per_asset", 0.0))
    max_w: float = float(active_constraints.get("max_weight_per_asset", 0.25))

    # CRITICAL FIX: Extract n_simulations for CVaR, defaulting to 5000 for robust KDE smoothing
    n_simulations: int = int(active_constraints.get("cvar_n_simulations", 5000))

    # Define the robust starting point (Equal Weight) for iterative solvers
    w0: np.ndarray = np.full(n_assets, 1.0 / n_assets, dtype=np.float64)

    # Initialize variables to hold the final results
    final_weights: np.ndarray = w0.copy()
    objective_val: float = 0.0
    diagnostics: Dict[str, Any] = {"solver": "SLSQP", "success": True, "message": "Optimization successful"}

    # Set the random seed for stochastic methods (e.g., Resampled MVO, HRP)
    np.random.seed(optimizer_seed)

    # --- Define Objective Functions for the Registry ---

    def _obj_gmv(w: np.ndarray) -> float:
        # Objective: Minimize portfolio variance (w^T * Sigma * w)
        return float(w.T @ sigma @ w)

    def _obj_max_sharpe(w: np.ndarray) -> float:
        # Objective: Maximize Sharpe Ratio (Minimize negative Sharpe)
        port_return: float = float(w.T @ mu)
        port_vol: float = float(np.sqrt(w.T @ sigma @ w))
        # Guard against division by zero
        if port_vol < 1e-8:
            return 0.0
        return -((port_return - rf) / port_vol)

    def _obj_erc(w: np.ndarray) -> float:
        # Objective: Equal Risk Contribution (Minimize sum of squared differences from 1/N)
        port_var: float = float(w.T @ sigma @ w)
        # Guard against division by zero
        if port_var < 1e-8:
            return 1e6
        # Marginal Risk Contribution
        mrc: np.ndarray = sigma @ w
        # Total Risk Contribution
        trc: np.ndarray = w * mrc
        # Percentage Risk Contribution
        prc: np.ndarray = trc / port_var
        target: float = 1.0 / n_assets
        return float(np.sum((prc - target)**2))

    def _obj_max_div(w: np.ndarray) -> float:
        # Objective: Maximize Diversification Ratio (Minimize negative ratio)
        asset_vols: np.ndarray = np.sqrt(np.diag(sigma))
        weighted_vols: float = float(w.T @ asset_vols)
        port_vol: float = float(np.sqrt(w.T @ sigma @ w))
        # Guard against division by zero
        if port_vol < 1e-8:
            return 0.0
        return -(weighted_vols / port_vol)

    # --- Dynamic Registry ---
    # Maps method slugs to a tuple: (objective_callable, requires_returns_matrix, is_heuristic)
    registry: Dict[str, Tuple[Optional[Callable], bool, bool]] = {
        "equal_weight": (None, False, True),
        "iv_weight": (None, False, True),
        "gmv": (_obj_gmv, False, False),
        "max_sharpe": (_obj_max_sharpe, False, False),
        "bl_posterior": (_obj_max_sharpe, False, False), # Uses BL mu passed by orchestrator
        "regime_conditional": (_obj_max_sharpe, False, False), # Uses Regime mu
        "erc": (_obj_erc, False, False),
        "risk_parity": (_obj_erc, False, False),
        "max_diversification": (_obj_max_div, False, False),
        "hrp": (None, False, True), # Handled via custom function
        "min_cvar": (None, True, True) # Handled via LP
    }

    # Check if the requested method is in the primary registry
    if method not in registry:
        # Fallback for advanced methods not fully implemented in this snippet
        logger.warning(f"Method '{method}' not in primary registry. Falling back to ERC proxy.")
        method_to_run = "erc"
        diagnostics["proxy_used"] = "erc"
    else:
        method_to_run = method

    # Extract the routing parameters for the selected method
    obj_func, req_returns, is_heuristic = registry[method_to_run]

    try:
        # Guard: Ensure returns matrix is provided if the method requires it
        if req_returns and returns_matrix is None:
            raise ValueError(f"returns_matrix is required for method '{method_to_run}'.")

        if is_heuristic:
            # Route to specific heuristic or custom solver implementations
            if method_to_run == "equal_weight":
                final_weights = w0
                diagnostics["solver"] = "Analytic"
            elif method_to_run == "iv_weight":
                asset_vols = np.sqrt(np.diag(sigma))
                inv_vols = 1.0 / np.maximum(asset_vols, 1e-8)
                final_weights = inv_vols / np.sum(inv_vols)
                diagnostics["solver"] = "Analytic"
            elif method_to_run == "hrp":
                # Call the custom HRP allocation function
                final_weights = _hrp_allocation(sigma, min_w, max_w)
                diagnostics["solver"] = "Hierarchical_Clustering"
            elif method_to_run == "min_cvar":
                # CRITICAL FIX: Pass n_simulations to the LP optimizer
                final_weights = _cvar_lp_optimization(
                    returns_matrix,
                    alpha=0.95,
                    min_w=min_w,
                    max_w=max_w,
                    n_simulations=n_simulations
                )
                diagnostics["solver"] = "Highs_LP"
        else:
            # Standard Non-Linear Programming (NLP) Optimization via SLSQP
            # Define the box bounds: w_i \in [min_w, max_w]
            bounds = Bounds(
                lb=np.full(n_assets, min_w, dtype=np.float64),
                ub=np.full(n_assets, max_w, dtype=np.float64)
            )

            # Define the budget equality constraint: \sum w_i = 1.0
            A_eq = np.ones((1, n_assets), dtype=np.float64)
            budget_constraint = LinearConstraint(
                A=A_eq,
                lb=np.array([1.0], dtype=np.float64),
                ub=np.array([1.0], dtype=np.float64)
            )

            # Execute the SLSQP solver
            opt_result: OptimizeResult = minimize(
                fun=obj_func,
                x0=w0,
                method="SLSQP",
                bounds=bounds,
                constraints=[budget_constraint],
                options={"ftol": 1e-9, "maxiter": 1000}
            )

            # Check solver success
            if opt_result.success:
                final_weights = opt_result.x
                objective_val = float(opt_result.fun)
                diagnostics["message"] = opt_result.message
            else:
                raise ValueError(f"SLSQP failed: {opt_result.message}")

    except Exception as exc:
        # Fail-safe: If optimization fails, log the error and fall back to Equal Weight
        logger.error(f"Optimization failed for '{method}': {exc}. Falling back to Equal Weight.")
        final_weights = w0
        diagnostics["success"] = False
        diagnostics["message"] = f"FAILED: {str(exc)}. Fallback to EW applied."

    # Apply rigorous Euclidean projection to guarantee strict JSON schema compliance
    # This eliminates floating-point drift (e.g., weights summing to 0.999999999)
    projected_weights: np.ndarray = _apply_post_optimization_projection(final_weights, min_w, max_w)

    # Log the successful completion of the optimization
    logger.debug(f"Optimization complete for '{method}'. Sum of weights: {np.sum(projected_weights):.6f}")

    # Return the structured results dictionary
    return {
        "method": method,
        "weights": projected_weights.tolist(),
        "objective_value": objective_val,
        "diagnostics": diagnostics
    }

# -------------------------------------------------------------------------------------------------------------------------------
# Task 25, Step 3: Persist each candidate (pc_weights.json and pc_report.md)
# -------------------------------------------------------------------------------------------------------------------------------

def write_pc_weights_json(
    method: str,
    weights: List[float],
    rationale: str,
    artifact_dir: Path
) -> str:
    """
    Serializes and persists the optimized portfolio weights to pc_weights.json.
    Enforces strict schema compliance (sum to 1.0, non-negative, 18 elements).

    Parameters
    ----------
    method : str
        The canonical slug of the PC method.
    weights : List[float]
        The 18-element list of optimized weights.
    rationale : str
        The LLM-generated rationale string.
    artifact_dir : Path
        The base directory for PC agents (e.g., run_outputs/pc_agents/).

    Returns
    -------
    str
        The absolute path to the written JSON file.

    Raises
    ------
    ValueError
        If the weights list does not contain exactly 18 elements.
    OSError
        If the file cannot be written.
    """
    if len(weights) != 18:
        raise ValueError(f"Weights list must contain exactly 18 elements, got {len(weights)}.")

    # Calculate the sum for the schema's sum_check field
    sum_check: float = float(sum(weights))

    # Construct the dictionary conforming to pc_weights.schema.json
    payload: Dict[str, Any] = {
        "method": method,
        "weights": weights,
        "sum_check": sum_check,
        "rationale": rationale
    }

    # Ensure the method-specific subdirectory exists
    method_dir: Path = artifact_dir / method
    method_dir.mkdir(parents=True, exist_ok=True)

    output_path: Path = method_dir / "pc_weights.json"

    try:
        with open(output_path, "w", encoding="utf-8") as f:
            json.dump(payload, f, indent=4)
    except OSError as exc:
        raise OSError(f"Failed to write pc_weights.json for '{method}': {exc}") from exc

    logger.info(f"Successfully persisted pc_weights.json for '{method}'.")
    return str(output_path.resolve())


def write_pc_report_md(
    method: str,
    weights: List[float],
    rationale: str,
    artifact_dir: Path,
    benchmark_weights: Optional[List[float]] = None,
    macro_view: Optional[Dict[str, Any]] = None,
    asset_class_labels: Optional[List[str]] = None
) -> str:
    """
    Formats and persists the human-readable portfolio construction report to pc_report.md.

    Parameters
    ----------
    method : str
        The canonical slug of the PC method.
    weights : List[float]
        The 18-element list of optimized weights.
    rationale : str
        The LLM-generated rationale string.
    artifact_dir : Path
        The base directory for PC agents.
    benchmark_weights : Optional[List[float]], optional
        The 18-element benchmark weights for relative exposure calculation.
    macro_view : Optional[Dict[str, Any]], optional
        The macro view dictionary for context.
    asset_class_labels : Optional[List[str]], optional
        The canonical names of the 18 assets.

    Returns
    -------
    str
        The absolute path to the written Markdown file.

    Raises
    ------
    OSError
        If the file cannot be written.
    """
    # Provide fallback labels if none are provided
    if asset_class_labels is None:
        asset_class_labels = [f"Asset_{i+1}" for i in range(18)]

    # Construct the markdown content
    md_lines: List[str] = [
        f"# Portfolio Construction Report: {method.replace('_', ' ').title()}",
        "",
        "## 1. Objective and Rationale",
        rationale,
        "",
        "## 2. Absolute Allocations",
        "| Asset Class | Weight |",
        "| :--- | :--- |"
    ]

    for label, w in zip(asset_class_labels, weights):
        md_lines.append(f"| {label} | {w * 100:.2f}% |")

    md_lines.append("")

    # Add relative exposures if benchmark weights are provided
    if benchmark_weights is not None and len(benchmark_weights) == 18:
        md_lines.extend([
            "## 3. Active Exposures (vs Benchmark)",
            "| Asset Class | Active Weight |",
            "| :--- | :--- |"
        ])
        for label, w, bw in zip(asset_class_labels, weights, benchmark_weights):
            active_w = w - bw
            if abs(active_w) > 0.001: # Only show meaningful active bets
                md_lines.append(f"| {label} | {active_w * 100:.2f}% |")
        md_lines.append("")

    if macro_view is not None:
        regime = macro_view.get("regime", "Unknown")
        md_lines.extend([
            "## 4. Macro Context",
            f"Portfolio constructed under **{regime}** regime conditions.",
            ""
        ])

    md_content: str = "\n".join(md_lines)

    # Ensure the method-specific subdirectory exists
    method_dir: Path = artifact_dir / method
    method_dir.mkdir(parents=True, exist_ok=True)

    output_path: Path = method_dir / "pc_report.md"

    try:
        with open(output_path, "w", encoding="utf-8") as f:
            f.write(md_content)
    except OSError as exc:
        raise OSError(f"Failed to write pc_report.md for '{method}': {exc}") from exc

    logger.info(f"Successfully persisted pc_report.md for '{method}'.")
    return str(output_path.resolve())

# -------------------------------------------------------------------------------------------------------------------------------
# Task 25, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_25_construct_pc_candidates(
    sigma: np.ndarray,
    mu_dict: Dict[str, np.ndarray],
    rf: float,
    benchmark_weights: np.ndarray,
    returns_matrix: np.ndarray,
    config: Dict[str, Any],
    run_output_dir: Path
) -> Dict[str, str]:
    """
    Orchestrates the computational generation of all 20 canonical PC candidate
    portfolios. Iterates through the registry, executes the optimizers, and
    persists the artifacts.

    Note: In the full architecture, this computational step is invoked by the
    LLM agents during Phase A. This orchestrator simulates that execution block
    for the deterministic mathematical layer.

    Parameters
    ----------
    sigma : np.ndarray
        The 18x18 annualized covariance matrix.
    mu_dict : Dict[str, np.ndarray]
        Mapping from method slug to 18-element expected-return vector.
    rf : float
        The annualized risk-free rate.
    benchmark_weights : np.ndarray
        The 18-element benchmark weight vector.
    returns_matrix : np.ndarray
        The T x 18 returns matrix.
    config : Dict[str, Any]
        The validated study configuration dictionary.
    run_output_dir : Path
        The root directory for the current run's artifacts.

    Returns
    -------
    Dict[str, str]
        A dictionary mapping method slugs to their generated pc_weights.json paths.

    Raises
    ------
    KeyError
        If required configuration keys are missing.
    """
    logger.info("Starting Task 25: Construct PC candidate portfolios (20+ methods) under IPS constraints.")

    try:
        pc_registry: Dict[str, Any] = config["REGISTRIES"]["PC_AGENTS"]["method_configs"]
        ips_constraints: Dict[str, Any] = config["IPS_GOVERNANCE"]["IMPLEMENTATION_CONSTRAINTS"]

        # Extract canonical asset labels for reporting
        asset_classes_dict: Dict[str, List[str]] = config["IPS_GOVERNANCE"]["ASSET_CLASSES"]
        asset_class_labels: List[str] = []
        for category in ["equity", "fixed_income", "real_assets", "cash"]:
            asset_class_labels.extend(asset_classes_dict.get(category, []))

    except KeyError as exc:
        raise KeyError(f"Configuration is missing required keys for Task 25: {exc}") from exc

    pc_artifact_dir: Path = run_output_dir / "pc_agents"
    generated_artifacts: Dict[str, str] = {}

    # Iterate through all registered methods
    for method_slug, method_config in pc_registry.items():
        # Skip special agents not part of the canonical 20
        if method_slug in ["researcher", "adversarial_diversifier"]:
            continue

        logger.info(f"Processing PC candidate: {method_slug}")

        # Retrieve the correct mu vector if the method uses CMAs
        uses_cmas: bool = method_config.get("uses_cmas", False)
        mu_vector: np.ndarray = mu_dict.get(method_slug, np.zeros(18, dtype=np.float64)) if uses_cmas else np.zeros(18, dtype=np.float64)

        # Step 1 & 2: Execute the optimizer
        opt_result: Dict[str, Any] = run_optimizer(
            method=method_slug,
            sigma=sigma,
            mu=mu_vector,
            rf=rf,
            benchmark_weights=benchmark_weights,
            constraints=ips_constraints,
            returns_matrix=returns_matrix
        )

        weights: List[float] = opt_result["weights"]

        # Generate a deterministic rationale string
        rationale: str = (
            f"Portfolio optimized using {method_slug} objective. "
            f"Solver status: {opt_result['diagnostics']['success']}. "
            f"Objective value: {opt_result['objective_value']:.6f}."
        )

        # Step 3: Persist the artifacts
        json_path: str = write_pc_weights_json(
            method=method_slug,
            weights=weights,
            rationale=rationale,
            artifact_dir=pc_artifact_dir
        )

        write_pc_report_md(
            method=method_slug,
            weights=weights,
            rationale=rationale,
            artifact_dir=pc_artifact_dir,
            benchmark_weights=benchmark_weights.tolist(),
            asset_class_labels=asset_class_labels
        )

        generated_artifacts[method_slug] = json_path

    logger.info(f"Task 25 completed successfully. Generated {len(generated_artifacts)} PC candidate portfolios.")
    return generated_artifacts
````

````python
# Task 26 — Execute adversarial diversifier (must run after centroid exists)

# ==============================================================================
# Task 26: Execute adversarial diversifier (must run after centroid exists)
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 26, Step 1: Compute centroid of other PC portfolios
# -------------------------------------------------------------------------------------------------------------------------------

def compute_ensemble_centroid(
    other_pc_weights_dir: Path,
    n_methods: int = 20,
    method_slugs: Optional[List[str]] = None
) -> Dict[str, Any]:
    """
    Computes the equal-weighted centroid of all non-adversarial PC portfolios.
    Enforces a hard synchronization gate: it will only compute the centroid if
    exactly `n_methods` valid pc_weights.json artifacts are present.

    Equation: \bar{w} = \frac{1}{K}\sum_{k=1}^K w^{(k)}

    Parameters
    ----------
    other_pc_weights_dir : Path
        The base directory containing the subdirectories for each PC method
        (e.g., run_outputs/pc_agents/).
    n_methods : int, optional
        The exact number of non-adversarial methods expected. Defaults to 20.
    method_slugs : Optional[List[str]], optional
        An explicit list of method slugs to include. If None, it scans the directory.

    Returns
    -------
    Dict[str, Any]
        A dictionary containing the 18-element centroid weight vector and metadata.

    Raises
    ------
    RuntimeError
        If the number of found artifacts does not exactly match `n_methods`.
    ValueError
        If any JSON artifact is unparseable or contains invalid weight arrays.
    """
    logger.info(f"Computing ensemble centroid. Expecting exactly {n_methods} non-adversarial portfolios.")

    # Identify the target JSON files
    target_files: List[Path] = []
    if method_slugs is not None:
        for slug in method_slugs:
            if slug == "adversarial_diversifier" or slug == "researcher":
                continue
            file_path = other_pc_weights_dir / slug / "pc_weights.json"
            if file_path.exists():
                target_files.append(file_path)
    else:
        # Scan the directory, excluding the adversarial and researcher directories
        for p in other_pc_weights_dir.glob("*/pc_weights.json"):
            if p.parent.name not in ["adversarial_diversifier", "researcher"]:
                target_files.append(p)

    # Hard synchronization gate: ensure exactly K portfolios exist
    found_count: int = len(target_files)
    if found_count != n_methods:
        raise RuntimeError(
            f"Synchronization Gate Failed: Expected exactly {n_methods} non-adversarial "
            f"pc_weights.json artifacts, but found {found_count}. The Adversarial Diversifier "
            f"cannot execute until Phase A is fully complete."
        )

    # Load and stack the weight vectors
    weights_list: List[np.ndarray] = []
    included_slugs: List[str] = []

    for file_path in target_files:
        try:
            with open(file_path, "r", encoding="utf-8") as f:
                data: Dict[str, Any] = json.load(f)

            w_array: np.ndarray = np.array(data["weights"], dtype=np.float64)

            # Validate dimensionality
            if w_array.shape != (18,):
                raise ValueError(f"Weight array in {file_path.parent.name} has shape {w_array.shape}, expected (18,).")

            weights_list.append(w_array)
            included_slugs.append(data.get("method", file_path.parent.name))

        except (json.JSONDecodeError, KeyError) as exc:
            raise ValueError(f"Failed to parse or extract weights from {file_path}: {exc}") from exc

    # Stack into a K x 18 matrix
    weights_matrix: np.ndarray = np.vstack(weights_list)

    # Compute the column-wise mean to find the centroid
    centroid_weights: np.ndarray = np.mean(weights_matrix, axis=0)

    # Validate that the centroid sums to 1.0
    centroid_sum: float = float(np.sum(centroid_weights))
    if abs(centroid_sum - 1.0) > 1e-6:
        # Renormalize if floating point drift occurred during averaging
        centroid_weights = centroid_weights / centroid_sum

    logger.info(f"Successfully computed ensemble centroid from {found_count} portfolios.")

    return {
        "centroid_weights": centroid_weights.tolist(),
        "n_portfolios_used": found_count,
        "method_slugs_included": included_slugs
    }

# -------------------------------------------------------------------------------------------------------------------------------
# Task 26, Step 2: Compute Sharpe floor
# -------------------------------------------------------------------------------------------------------------------------------

def check_sharpe_floor(
    max_sharpe_diagnostics_path: Path,
    returns_matrix: np.ndarray,
    rf: float,
    achieved_sharpe: Optional[float] = None
) -> Dict[str, Any]:
    """
    Computes the Sharpe floor (75% of the Max Sharpe portfolio's backtest Sharpe)
    for the adversarial diversifier constraint. Bounds the floor at 0.0 to prevent
    optimizer divergence in severe bear market simulations.

    Equation: SR_{floor} = \max(0.0, 0.75 \cdot SR(w_{\max SR}))

    Parameters
    ----------
    max_sharpe_diagnostics_path : Path
        The path to the pc_weights.json artifact for the max_sharpe method.
    returns_matrix : np.ndarray
        The T x 18 monthly returns matrix.
    rf : float
        The annualized risk-free rate.
    achieved_sharpe : Optional[float], optional
        The Sharpe ratio achieved by the adversarial optimizer, used to verify
        if the floor was met post-optimization. Defaults to None.

    Returns
    -------
    Dict[str, Any]
        A dictionary containing the computed floor, the max Sharpe value, and
        a boolean indicating if the floor was met (if achieved_sharpe is provided).

    Raises
    ------
    FileNotFoundError
        If the max_sharpe artifact does not exist.
    ValueError
        If the artifact is unparseable.
    """
    logger.info("Computing Sharpe floor for Adversarial Diversifier.")

    # Verify the artifact exists
    if not max_sharpe_diagnostics_path.exists():
        raise FileNotFoundError(
            f"Max Sharpe artifact not found at {max_sharpe_diagnostics_path}. "
            f"Cannot compute Sharpe floor."
        )

    # Load the Max Sharpe weights
    try:
        with open(max_sharpe_diagnostics_path, "r", encoding="utf-8") as f:
            data: Dict[str, Any] = json.load(f)
        w_max_sr: np.ndarray = np.array(data["weights"], dtype=np.float64)
    except (json.JSONDecodeError, KeyError) as exc:
        raise ValueError(f"Failed to parse Max Sharpe weights from {max_sharpe_diagnostics_path}: {exc}") from exc

    # Compute the backtest Sharpe ratio for the Max Sharpe portfolio
    # Note: compute_backtest_sharpe expects a list[float]
    max_sharpe_value: float = compute_backtest_sharpe(
        weights=w_max_sr.tolist(),
        returns_matrix=returns_matrix,
        rf=rf
    )

    # Compute the floor: 75% of the max Sharpe, bounded at 0.0
    sharpe_floor: float = max(0.0, 0.75 * max_sharpe_value)

    logger.info(f"Computed Sharpe floor: {sharpe_floor:.4f} (75% of Max Sharpe: {max_sharpe_value:.4f})")

    result: Dict[str, Any] = {
        "sharpe_floor": sharpe_floor,
        "max_sharpe_value": max_sharpe_value,
        "floor_met": None
    }

    # If an achieved Sharpe is provided, verify if it meets the floor
    if achieved_sharpe is not None:
        # Allow a tiny floating point tolerance
        floor_met: bool = achieved_sharpe >= (sharpe_floor - 1e-6)
        result["floor_met"] = floor_met
        if not floor_met:
            logger.warning(f"Adversarial portfolio failed to meet Sharpe floor. Achieved: {achieved_sharpe:.4f}, Floor: {sharpe_floor:.4f}")

    return result

# -------------------------------------------------------------------------------------------------------------------------------
# Task 26, Step 3: Solve adversarial objective
# -------------------------------------------------------------------------------------------------------------------------------

def run_adversarial_diversifier_optimizer(
    centroid_weights: List[float],
    sharpe_floor: float,
    sigma: np.ndarray,
    mu: np.ndarray,
    rf: float,
    benchmark_weights: np.ndarray,
    constraints: Dict[str, Any],
    optimizer_seed: int = 24680
) -> Dict[str, Any]:
    """
    Solves the adversarial diversifier optimization problem by maximizing tracking
    variance against the centroid subject to a Sharpe ratio floor and IPS constraints.

    CRITICAL REMEDIATION: Replaces the brittle Differential Evolution (DE) solver
    with a mathematically rigorous Sequential Convex Programming (SCP) algorithm.
    By linearizing the convex tracking variance objective at each iteration, we
    solve a sequence of strictly convex sub-problems using SLSQP, guaranteeing
    monotonic convergence without warping the search space.

    Objective: \max_w (w - \bar{w})^\top \Sigma (w - \bar{w})
    Subject to: w^\top \mu - r_f - SR_{floor} \sqrt{w^\top \Sigma w} \ge 0

    Parameters
    ----------
    centroid_weights : List[float]
        The 18-element centroid weight vector (\bar{w}).
    sharpe_floor : float
        The minimum acceptable annualized Sharpe ratio.
    sigma : np.ndarray
        The 18x18 annualized covariance matrix.
    mu : np.ndarray
        The 18x1 expected return vector (typically the Max Sharpe CMAs).
    rf : float
        The annualized risk-free rate.
    benchmark_weights : np.ndarray
        The 18x1 benchmark weight vector (for IPS TE constraint).
    constraints : Dict[str, Any]
        The IPS implementation constraints dictionary.
    optimizer_seed : int, optional
        Seed for stochastic elements (unused in deterministic SCP, kept for signature).

    Returns
    -------
    Dict[str, Any]
        A dictionary containing the optimized weights, tracking variance, achieved Sharpe, and diagnostics.

    Raises
    ------
    RuntimeError
        If the SCP solver fails to converge to a feasible solution.
    """
    logger.info("Executing Sequential Convex Programming (SCP) for Adversarial Diversifier.")

    n_assets: int = sigma.shape[0]
    w_bar: np.ndarray = np.array(centroid_weights, dtype=np.float64)

    min_w: float = float(constraints.get("min_weight_per_asset", 0.0))
    max_w: float = float(constraints.get("max_weight_per_asset", 0.25))

    # Define the Box Bounds
    bounds = Bounds(
        lb=np.full(n_assets, min_w, dtype=np.float64),
        ub=np.full(n_assets, max_w, dtype=np.float64)
    )

    # Define the Budget Equality Constraint: \sum w_i = 1.0
    A_eq = np.ones((1, n_assets), dtype=np.float64)
    budget_constraint = LinearConstraint(
        A=A_eq,
        lb=np.array([1.0], dtype=np.float64),
        ub=np.array([1.0], dtype=np.float64)
    )

    # Define the exact Second-Order Cone (SOC) Sharpe Floor Constraint
    # g(w) = w^T \mu - r_f - SR_{floor} \sqrt{w^T \Sigma w} \ge 0
    def _sharpe_constraint_fun(w: np.ndarray) -> float:
        port_return = float(w.T @ mu)
        port_vol = float(np.sqrt(max(w.T @ sigma @ w, 1e-12)))
        return port_return - rf - (sharpe_floor * port_vol)

    # Analytical Jacobian of the Sharpe constraint for rapid SLSQP convergence
    # \nabla g(w) = \mu - SR_{floor} \frac{\Sigma w}{\sqrt{w^T \Sigma w}}
    def _sharpe_constraint_jac(w: np.ndarray) -> np.ndarray:
        port_vol = float(np.sqrt(max(w.T @ sigma @ w, 1e-12)))
        return mu - (sharpe_floor * (sigma @ w) / port_vol)

    sharpe_nlc = NonlinearConstraint(
        fun=_sharpe_constraint_fun,
        lb=0.0,
        ub=np.inf,
        jac=_sharpe_constraint_jac
    )

    # Initialize SCP at the Max Sharpe portfolio (or equal weight if mu is zero)
    # This ensures immediate feasibility for the Sharpe floor constraint
    w_k: np.ndarray = np.full(n_assets, 1.0 / n_assets, dtype=np.float64)
    if np.any(mu != 0):
        # Quick approximation of Max Sharpe for initialization
        inv_sigma_mu = np.linalg.solve(sigma, mu)
        w_k = inv_sigma_mu / np.sum(inv_sigma_mu)
        w_k = _apply_post_optimization_projection(w_k, min_w, max_w)

    # SCP Parameters
    max_scp_iter: int = 20
    scp_tol: float = 1e-5
    iteration: int = 0
    converged: bool = False

    # Execute the Sequential Convex Programming loop
    while iteration < max_scp_iter:
        # Linearize the objective at the current guess w_k
        # f(w) = (w - \bar{w})^T \Sigma (w - \bar{w})
        # \nabla f(w_k) = 2 \Sigma (w_k - \bar{w})
        # We want to maximize \nabla f(w_k)^T w, so we minimize - \nabla f(w_k)^T w
        grad_f_wk: np.ndarray = 2.0 * sigma @ (w_k - w_bar)

        def _linearized_objective(w: np.ndarray) -> float:
            return float(-grad_f_wk.T @ w)

        def _linearized_jacobian(w: np.ndarray) -> np.ndarray:
            return -grad_f_wk

        # Solve the strictly convex sub-problem using SLSQP
        res = minimize(
            fun=_linearized_objective,
            x0=w_k,
            jac=_linearized_jacobian,
            method="SLSQP",
            bounds=bounds,
            constraints=[budget_constraint, sharpe_nlc],
            options={"ftol": 1e-8, "maxiter": 200, "disp": False}
        )

        if not res.success:
            logger.warning(f"SCP sub-problem failed at iteration {iteration}: {res.message}")
            break

        w_next: np.ndarray = res.x

        # Check for convergence (L2 norm of the step)
        step_size: float = float(np.linalg.norm(w_next - w_k))
        if step_size < scp_tol:
            w_k = w_next
            converged = True
            break

        # Update the current guess
        w_k = w_next
        iteration += 1

    if not converged and iteration == max_scp_iter:
        logger.warning("SCP reached maximum iterations without strict convergence. Using best found weights.")

    # Apply rigorous Euclidean projection to guarantee strict JSON schema compliance
    projected_weights: np.ndarray = _apply_post_optimization_projection(w_k, min_w, max_w)

    # Compute final achieved metrics for reporting
    achieved_tv: float = float((projected_weights - w_bar).T @ sigma @ (projected_weights - w_bar))
    final_port_vol: float = float(np.sqrt(max(projected_weights.T @ sigma @ projected_weights, 1e-12)))
    final_achieved_sharpe: float = float((projected_weights.T @ mu - rf) / final_port_vol)

    # Final feasibility check
    if final_achieved_sharpe < sharpe_floor - 1e-4:
        raise RuntimeError(
            f"Adversarial SCP failed to find a feasible solution meeting the Sharpe floor. "
            f"Achieved: {final_achieved_sharpe:.4f}, Required: {sharpe_floor:.4f}. Pipeline halted."
        )

    logger.info(f"Adversarial SCP optimization successful. Max tracking variance: {achieved_tv:.6f}")

    return {
        "weights": projected_weights.tolist(),
        "tracking_variance_vs_centroid": achieved_tv,
        "achieved_sharpe": final_achieved_sharpe,
        "diagnostics": {"solver": "Sequential_Convex_Programming", "success": True, "iterations": iteration}
    }

# -------------------------------------------------------------------------------------------------------------------------------
# Task 26, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_26_adversarial_diversifier(
    sigma: np.ndarray,
    mu_dict: Dict[str, np.ndarray],
    rf: float,
    benchmark_weights: np.ndarray,
    returns_matrix: np.ndarray,
    config: Dict[str, Any],
    run_output_dir: Path
) -> str:
    """
    Orchestrates the computational generation of the Adversarial Diversifier portfolio.
    Computes the centroid, establishes the Sharpe floor, executes the global
    optimizer, and persists the artifacts.

    Parameters
    ----------
    sigma : np.ndarray
        The 18x18 annualized covariance matrix.
    mu_dict : Dict[str, np.ndarray]
        Mapping from method slug to 18-element expected-return vector.
    rf : float
        The annualized risk-free rate.
    benchmark_weights : np.ndarray
        The 18-element benchmark weight vector.
    returns_matrix : np.ndarray
        The T x 18 returns matrix.
    config : Dict[str, Any]
        The validated study configuration dictionary.
    run_output_dir : Path
        The root directory for the current run's artifacts.

    Returns
    -------
    str
        The path to the generated pc_weights.json artifact.

    Raises
    ------
    KeyError
        If required configuration keys are missing.
    RuntimeError
        If the synchronization gate (20 portfolios) fails or optimization fails.
    """
    # Log the commencement of the Adversarial Diversifier task
    logger.info("Starting Task 26: Execute adversarial diversifier (must run after centroid exists).")

    # Extract configuration parameters
    try:
        ips_constraints: Dict[str, Any] = config["IPS_GOVERNANCE"]["IMPLEMENTATION_CONSTRAINTS"]
        optimizer_seed: int = config["RANDOM_SEEDS"]["optimizer_seed"]

        # Extract canonical asset labels for reporting
        asset_classes_dict: Dict[str, List[str]] = config["IPS_GOVERNANCE"]["ASSET_CLASSES"]
        asset_class_labels: List[str] = []
        for category in ["equity", "fixed_income", "real_assets", "cash"]:
            asset_class_labels.extend(asset_classes_dict.get(category, []))

    except KeyError as exc:
        # Raise a KeyError if the expected nested structures are missing
        raise KeyError(f"Configuration is missing required keys for Task 26: {exc}") from exc

    # Define the specific artifact directory for PC Agents
    pc_artifact_dir: Path = run_output_dir / "pc_agents"
    method_slug: str = "adversarial_diversifier"

    # Step 1: Compute centroid of other PC portfolios
    # This enforces the hard synchronization gate: exactly 20 portfolios must exist
    centroid_result: Dict[str, Any] = compute_ensemble_centroid(
        other_pc_weights_dir=pc_artifact_dir,
        n_methods=20
    )
    centroid_weights: List[float] = centroid_result["centroid_weights"]

    # Step 2: Compute Sharpe floor
    # The floor is 75% of the Max Sharpe portfolio's backtest Sharpe
    max_sharpe_path: Path = pc_artifact_dir / "max_sharpe" / "pc_weights.json"
    floor_result: Dict[str, Any] = check_sharpe_floor(
        max_sharpe_diagnostics_path=max_sharpe_path,
        returns_matrix=returns_matrix,
        rf=rf
    )
    sharpe_floor: float = floor_result["sharpe_floor"]

    # Extract the Max Sharpe mu vector for the adversarial constraint
    # If max_sharpe mu is missing, fall back to zeros (though this implies a pipeline failure upstream)
    mu_max_sharpe: np.ndarray = mu_dict.get("max_sharpe", np.zeros(18, dtype=np.float64))

    # Step 3: Solve adversarial objective
    # CRITICAL FIX: Removed n_starts=10 argument
    opt_result: Dict[str, Any] = run_adversarial_diversifier_optimizer(
        centroid_weights=centroid_weights,
        sharpe_floor=sharpe_floor,
        sigma=sigma,
        mu=mu_max_sharpe,
        rf=rf,
        benchmark_weights=benchmark_weights,
        constraints=ips_constraints,
        optimizer_seed=optimizer_seed
    )

    # Extract the optimized weights
    weights: List[float] = opt_result["weights"]

    # Generate a deterministic rationale string for the audit trail
    rationale: str = (
        f"Adversarial Diversifier: Maximized tracking variance vs centroid of {centroid_result['n_portfolios_used']} methods. "
        f"Solver status: {opt_result['diagnostics']['success']}. "
        f"Achieved Tracking Variance: {opt_result['tracking_variance_vs_centroid']:.6f}. "
        f"Achieved Sharpe: {opt_result['achieved_sharpe']:.4f} (Floor: {sharpe_floor:.4f})."
    )

    # Step 4: Persist the artifacts
    # Write the JSON weights file
    json_path: str = write_pc_weights_json(
        method=method_slug,
        weights=weights,
        rationale=rationale,
        artifact_dir=pc_artifact_dir
    )

    # Write the Markdown report
    write_pc_report_md(
        method=method_slug,
        weights=weights,
        rationale=rationale,
        artifact_dir=pc_artifact_dir,
        benchmark_weights=benchmark_weights.tolist(),
        asset_class_labels=asset_class_labels
    )

    # Log the successful completion of the orchestration task
    logger.info(f"Task 26 completed successfully. Adversarial Diversifier portfolio generated.")

    # Return the path to the generated JSON artifact
    return json_path
````

````python
# Task 27 — CRO risk reports (standardized diagnostics)

# ==============================================================================
# Task 27: CRO risk reports (standardized diagnostics)
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 27, Step 1: Implement and register CRO script tools; compute ex-ante vol and tracking error
# -------------------------------------------------------------------------------------------------------------------------------

def compute_ex_ante_vol(
    weights: List[float],
    sigma: np.ndarray
) -> float:
    """
    Computes the annualized ex-ante portfolio volatility using the quadratic form.

    Equation: \sigma_p = \sqrt{w^\top \Sigma w}

    Parameters
    ----------
    weights : List[float]
        The 18-element list of portfolio weights.
    sigma : np.ndarray
        The 18x18 annualized covariance matrix.

    Returns
    -------
    float
        The annualized ex-ante volatility.

    Raises
    ------
    ValueError
        If the weight vector length does not match the covariance matrix dimensions.
    """
    w_array: np.ndarray = np.array(weights, dtype=np.float64)

    if w_array.shape[0] != sigma.shape[0]:
        raise ValueError(f"Weight vector length ({w_array.shape[0]}) does not match covariance matrix dimension ({sigma.shape[0]}).")

    # Compute portfolio variance: w^T * Sigma * w
    port_var: float = float(w_array.T @ sigma @ w_array)

    # Guard against negative variance due to floating-point inaccuracies
    port_var = max(0.0, port_var)

    # Compute volatility
    port_vol: float = float(np.sqrt(port_var))

    logger.debug(f"Computed ex-ante volatility: {port_vol:.6f}")
    return port_vol


def compute_tracking_error(
    weights: List[float],
    benchmark_weights: np.ndarray,
    sigma: np.ndarray
) -> float:
    """
    Computes the annualized ex-ante tracking error relative to the benchmark.

    Equation: TE = \sqrt{(w - w_b)^\top \Sigma (w - w_b)}

    Parameters
    ----------
    weights : List[float]
        The 18-element list of portfolio weights.
    benchmark_weights : np.ndarray
        The 18-element benchmark weight vector.
    sigma : np.ndarray
        The 18x18 annualized covariance matrix.

    Returns
    -------
    float
        The annualized ex-ante tracking error.

    Raises
    ------
    ValueError
        If dimensionalities do not match.
    """
    w_array: np.ndarray = np.array(weights, dtype=np.float64)

    if w_array.shape != benchmark_weights.shape:
        raise ValueError("Portfolio weights and benchmark weights must have the same dimensions.")

    # Compute active weights: w - w_b
    active_weights: np.ndarray = w_array - benchmark_weights

    # Compute tracking variance: (w - w_b)^T * Sigma * (w - w_b)
    tracking_var: float = float(active_weights.T @ sigma @ active_weights)

    # Guard against negative variance
    tracking_var = max(0.0, tracking_var)

    # Compute tracking error
    tracking_error: float = float(np.sqrt(tracking_var))

    logger.debug(f"Computed ex-ante tracking error: {tracking_error:.6f}")
    return tracking_error

# -------------------------------------------------------------------------------------------------------------------------------
# Task 27, Step 2: Compute backtest Sharpe, MDD, and run factor regression
# -------------------------------------------------------------------------------------------------------------------------------

def compute_backtest_sharpe(
    weights: List[float],
    returns_matrix: np.ndarray,
    rf: float
) -> float:
    """
    Computes the annualized backtest Sharpe ratio for the portfolio.

    Equation: SR = \frac{\mathbb{E}[r_p - r_f]}{\sigma(r_p - r_f)}

    Parameters
    ----------
    weights : List[float]
        The 18-element list of portfolio weights.
    returns_matrix : np.ndarray
        The T x 18 monthly returns matrix.
    rf : float
        The annualized risk-free rate.

    Returns
    -------
    float
        The annualized backtest Sharpe ratio.
    """
    w_array: np.ndarray = np.array(weights, dtype=np.float64)

    # Compute historical portfolio returns: R_p = R * w
    port_returns: np.ndarray = returns_matrix @ w_array

    # De-annualize the risk-free rate for monthly excess return calculation
    rf_monthly: float = (1.0 + rf) ** (1.0 / 12.0) - 1.0

    # Compute excess returns
    excess_returns: np.ndarray = port_returns - rf_monthly

    # Compute mean and standard deviation of monthly excess returns
    mean_excess_mo: float = float(np.mean(excess_returns))
    std_excess_mo: float = float(np.std(excess_returns, ddof=1))

    # Guard against zero volatility
    if std_excess_mo < 1e-8:
        return 0.0

    # Annualize the metrics
    mean_excess_ann: float = mean_excess_mo * 12.0
    std_excess_ann: float = std_excess_mo * np.sqrt(12.0)

    # Compute Sharpe ratio
    sharpe: float = mean_excess_ann / std_excess_ann

    logger.debug(f"Computed backtest Sharpe ratio: {sharpe:.4f}")
    return sharpe


def compute_mdd(
    weights: List[float],
    returns_matrix: np.ndarray
) -> float:
    """
    Computes the maximum drawdown of the portfolio over the backtest window.

    Equation: MDD = \min_t \left(\frac{V_t}{\max_{s \le t} V_s} - 1\right)

    Parameters
    ----------
    weights : List[float]
        The 18-element list of portfolio weights.
    returns_matrix : np.ndarray
        The T x 18 monthly returns matrix.

    Returns
    -------
    float
        The maximum drawdown (as a negative decimal).
    """
    w_array: np.ndarray = np.array(weights, dtype=np.float64)

    # Compute historical portfolio returns
    port_returns: np.ndarray = returns_matrix @ w_array

    # Compute cumulative portfolio value (V_t)
    # V_t = \prod_{s=1}^t (1 + r_{p,s})
    cum_returns: np.ndarray = np.cumprod(1.0 + port_returns)

    # Compute the running maximum (max_{s \le t} V_s)
    running_max: np.ndarray = np.maximum.accumulate(cum_returns)

    # Guard against division by zero (though V_s should be > 0)
    running_max = np.maximum(running_max, 1e-8)

    # Compute drawdowns
    drawdowns: np.ndarray = (cum_returns / running_max) - 1.0

    # Extract the maximum drawdown (minimum value)
    mdd: float = float(np.min(drawdowns))

    logger.debug(f"Computed maximum drawdown: {mdd:.4f}")
    return mdd


def run_factor_regression(
    weights: List[float],
    returns_matrix: np.ndarray,
    returns_index: pd.DatetimeIndex,
    factors_df: pd.DataFrame,
    rf: float
) -> Dict[str, float]:
    """
    Fits the Fama-French 3-factor model to the portfolio's excess returns using OLS.

    CRITICAL REMEDIATION: Implements a strict variance floor check for `ss_tot`.
    If the portfolio is a constant-return asset (e.g., 100% Cash), `ss_tot` approaches
    zero, causing a divide-by-zero error or massive negative R-squared. This guard
    deterministically sets R-squared and betas to 0.0, preventing numerical explosion.

    Equation: r_{p,t} - r_{f,t} = \alpha + \beta_M MKT_t + \beta_S SMB_t + \beta_H HML_t + \varepsilon_t

    Parameters
    ----------
    weights : List[float]
        The 18-element list of portfolio weights.
    returns_matrix : np.ndarray
        The T x 18 monthly returns matrix.
    returns_index : pd.DatetimeIndex
        The temporal index corresponding to the returns_matrix.
    factors_df : pd.DataFrame
        The monthly Fama-French factor returns dataframe.
    rf : float
        The annualized risk-free rate (used if factors_df lacks an 'rf' column).

    Returns
    -------
    Dict[str, float]
        A dictionary containing the annualized alpha, betas, and R-squared.

    Raises
    ------
    ValueError
        If the temporal alignment results in an empty dataframe or if required
        factor columns are missing.
    """
    w_array: np.ndarray = np.array(weights, dtype=np.float64)

    # Compute historical portfolio returns
    port_returns_arr: np.ndarray = returns_matrix @ w_array

    # Create a pandas Series for the portfolio returns to enable index alignment
    port_returns_series = pd.Series(port_returns_arr, index=returns_index, name="port_return")

    # Ensure factors_df has a timezone-naive index for safe comparison
    df_factors_clean = factors_df.copy()
    if df_factors_clean.index.tz is not None:
        df_factors_clean.index = df_factors_clean.index.tz_convert(None)

    # Strip timezone from returns_index if necessary
    if port_returns_series.index.tz is not None:
        port_returns_series.index = port_returns_series.index.tz_convert(None)

    # Perform explicit inner join on the DatetimeIndex to guarantee temporal alignment
    aligned_df = pd.concat([port_returns_series, df_factors_clean], axis=1, join="inner")

    if aligned_df.empty:
        raise ValueError("Temporal alignment failed: No overlapping dates between portfolio returns and factor data.")

    # Extract aligned portfolio returns
    aligned_port_returns: np.ndarray = aligned_df["port_return"].values

    # Extract risk-free rate from aligned factors_df if available, else use de-annualized rf
    if "rf" in aligned_df.columns:
        rf_series: np.ndarray = aligned_df["rf"].values
    else:
        rf_monthly: float = (1.0 + rf) ** (1.0 / 12.0) - 1.0
        rf_series = np.full_like(aligned_port_returns, rf_monthly)

    # Compute excess returns (y)
    y: np.ndarray = aligned_port_returns - rf_series

    # Extract factors (X)
    required_factors = ["mkt_rf", "smb", "hml"]
    for f in required_factors:
        if f not in aligned_df.columns:
            raise ValueError(f"factors_df is missing required column: '{f}'")

    X_raw: np.ndarray = aligned_df[required_factors].values

    # Compute Total Sum of Squares (SS_tot) for the variance floor check
    y_mean: float = float(np.mean(y))
    ss_tot: float = float(np.sum((y - y_mean)**2))

    # CRITICAL REMEDIATION: Variance Floor Guard
    # If the portfolio has effectively zero variance (e.g., 100% Cash), OLS is undefined.
    # We deterministically set R-squared and betas to 0.0, and alpha to the mean excess return.
    if ss_tot < 1e-8:
        logger.info("Zero-variance portfolio detected (e.g., Cash). Bypassing OLS regression.")
        alpha_ann: float = y_mean * 12.0
        return {
            "alpha": alpha_ann,
            "beta_M": 0.0,
            "beta_S": 0.0,
            "beta_H": 0.0,
            "r_squared": 0.0
        }

    # Add intercept term (column of ones) for OLS
    X: np.ndarray = np.column_stack((np.ones(len(y)), X_raw))

    # Perform OLS regression using SVD-based least squares for numerical stability
    # Returns: coefficients, residuals, rank, singular values
    coeffs, residuals, rank, s = np.linalg.lstsq(X, y, rcond=None)

    # Extract coefficients
    alpha_monthly: float = float(coeffs[0])
    beta_m: float = float(coeffs[1])
    beta_s: float = float(coeffs[2])
    beta_h: float = float(coeffs[3])

    # Annualize alpha
    alpha_ann = alpha_monthly * 12.0

    # Compute R-squared
    if len(residuals) > 0:
        ss_res: float = float(residuals[0])
        r_squared: float = 1.0 - (ss_res / ss_tot)
    else:
        # If residuals are empty (exact fit), R-squared is 1.0
        r_squared = 1.0

    # Guard against floating point noise pushing R-squared slightly outside [0, 1]
    r_squared = float(np.clip(r_squared, 0.0, 1.0))

    logger.debug(f"Factor regression complete. Alpha (ann): {alpha_ann:.4f}, R2: {r_squared:.4f}")

    return {
        "alpha": alpha_ann,
        "beta_M": beta_m,
        "beta_S": beta_s,
        "beta_H": beta_h,
        "r_squared": r_squared
    }

# -------------------------------------------------------------------------------------------------------------------------------
# Task 27, Step 3: Check IPS compliance and write artifacts
# -------------------------------------------------------------------------------------------------------------------------------

def check_ips_compliance(
    vol: float,
    te: float,
    mdd: float,
    constraints: Dict[str, Any]
) -> Dict[str, Any]:
    """
    Evaluates the computed risk metrics against the IPS constraints.

    Parameters
    ----------
    vol : float
        The ex-ante annualized volatility.
    te : float
        The ex-ante annualized tracking error.
    mdd : float
        The maximum drawdown.
    constraints : Dict[str, Any]
        The IPS constraints dictionary containing the limits.

    Returns
    -------
    Dict[str, Any]
        A dictionary containing the overall compliance boolean and specific flags.
    """
    # Extract limits from constraints
    # Note: In the full architecture, these are nested in IPS_GOVERNANCE.
    # We assume the orchestrator flattens them or passes the relevant sub-dicts.
    vol_min: float = float(constraints.get("volatility_band_min", 0.08))
    vol_max: float = float(constraints.get("volatility_band_max", 0.12))
    te_limit: float = float(constraints.get("tracking_error_limit", 0.06))
    mdd_limit: float = float(constraints.get("max_drawdown_limit", -0.25))

    # Tolerance for floating-point comparisons
    tol: float = 1e-6

    # Evaluate conditions
    vol_compliant: bool = (vol_min - tol) <= vol <= (vol_max + tol)
    te_compliant: bool = te <= (te_limit + tol)
    mdd_compliant: bool = mdd >= (mdd_limit - tol)

    # Overall compliance
    overall_compliant: bool = vol_compliant and te_compliant and mdd_compliant

    flags: Dict[str, bool] = {
        "volatility_compliant": vol_compliant,
        "tracking_error_compliant": te_compliant,
        "max_drawdown_compliant": mdd_compliant
    }

    logger.debug(f"IPS Compliance Check: {overall_compliant}. Flags: {flags}")

    return {
        "compliant": overall_compliant,
        "flags": flags
    }


def write_cro_report_json(
    candidate_id: str,
    metrics: Dict[str, float],
    compliance_flags: Dict[str, Any],
    artifact_dir: Path
) -> str:
    """
    Serializes and persists the CRO risk report to cro_report.json.
    Enforces strict schema compliance.

    Parameters
    ----------
    candidate_id : str
        The canonical slug of the PC method.
    metrics : Dict[str, float]
        The dictionary of computed risk metrics.
    compliance_flags : Dict[str, Any]
        The dictionary of IPS compliance flags.
    artifact_dir : Path
        The base directory for CRO reports (e.g., run_outputs/cro_reports/).

    Returns
    -------
    str
        The absolute path to the written JSON file.

    Raises
    ------
    OSError
        If the file cannot be written.
    """
    payload: Dict[str, Any] = {
        "candidate_id": candidate_id,
        "ips_compliant": compliance_flags["compliant"],
        "metrics": metrics,
        "compliance_flags": compliance_flags["flags"]
    }

    # Ensure the method-specific subdirectory exists
    method_dir: Path = artifact_dir / candidate_id
    method_dir.mkdir(parents=True, exist_ok=True)

    output_path: Path = method_dir / "cro_report.json"

    try:
        with open(output_path, "w", encoding="utf-8") as f:
            json.dump(payload, f, indent=4)
    except OSError as exc:
        raise OSError(f"Failed to write cro_report.json for '{candidate_id}': {exc}") from exc

    logger.info(f"Successfully persisted cro_report.json for '{candidate_id}'.")
    return str(output_path.resolve())


def write_cro_narrative_md(
    candidate_id: str,
    metrics: Dict[str, float],
    rationale: str,
    artifact_dir: Path
) -> str:
    """
    Formats and persists the human-readable CRO risk narrative to cro_narrative.md.

    Parameters
    ----------
    candidate_id : str
        The canonical slug of the PC method.
    metrics : Dict[str, float]
        The dictionary of computed risk metrics.
    rationale : str
        The LLM-generated narrative string.
    artifact_dir : Path
        The base directory for CRO reports.

    Returns
    -------
    str
        The absolute path to the written Markdown file.

    Raises
    ------
    OSError
        If the file cannot be written.
    """
    # Construct the markdown content
    md_lines: List[str] = [
        f"# CRO Risk Report: {candidate_id.replace('_', ' ').title()}",
        "",
        "## 1. Risk Narrative",
        rationale,
        "",
        "## 2. Quantitative Diagnostics",
        "| Metric | Value |",
        "| :--- | :--- |",
        f"| Ex-Ante Volatility | {metrics.get('ex_ante_vol', 0.0) * 100:.2f}% |",
        f"| Tracking Error | {metrics.get('tracking_error', 0.0) * 100:.2f}% |",
        f"| Backtest Sharpe | {metrics.get('backtest_sharpe', 0.0):.4f} |",
        f"| Max Drawdown | {metrics.get('max_drawdown', 0.0) * 100:.2f}% |",
        f"| Alpha (Ann.) | {metrics.get('alpha', 0.0) * 100:.2f}% |",
        f"| Beta (Market) | {metrics.get('beta_M', 0.0):.4f} |",
        f"| Beta (Size) | {metrics.get('beta_S', 0.0):.4f} |",
        f"| Beta (Value) | {metrics.get('beta_H', 0.0):.4f} |",
        f"| R-Squared | {metrics.get('r_squared', 0.0):.4f} |",
        ""
    ]

    md_content: str = "\n".join(md_lines)

    # Ensure the method-specific subdirectory exists
    method_dir: Path = artifact_dir / candidate_id
    method_dir.mkdir(parents=True, exist_ok=True)

    output_path: Path = method_dir / "cro_narrative.md"

    try:
        with open(output_path, "w", encoding="utf-8") as f:
            f.write(md_content)
    except OSError as exc:
        raise OSError(f"Failed to write cro_narrative.md for '{candidate_id}': {exc}") from exc

    logger.info(f"Successfully persisted cro_narrative.md for '{candidate_id}'.")
    return str(output_path.resolve())

# -------------------------------------------------------------------------------------------------------------------------------
# Task 27, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_27_cro_reports(
    sigma: np.ndarray,
    rf: float,
    benchmark_weights: np.ndarray,
    returns_matrix: np.ndarray,
    returns_index: pd.DatetimeIndex,
    factors_df: pd.DataFrame,
    config: Dict[str, Any],
    run_output_dir: Path
) -> Dict[str, str]:
    """
    Orchestrates the computational generation of the CRO risk reports for all 21
    candidate portfolios. Iterates through the generated pc_weights.json artifacts,
    computes the risk metrics, checks IPS compliance, and persists the reports.

    Parameters
    ----------
    sigma : np.ndarray
        The 18x18 annualized covariance matrix.
    rf : float
        The annualized risk-free rate.
    benchmark_weights : np.ndarray
        The 18-element benchmark weight vector.
    returns_matrix : np.ndarray
        The T x 18 returns matrix.
    returns_index : pd.DatetimeIndex
        The temporal index corresponding to the returns_matrix.
    factors_df : pd.DataFrame
        The monthly Fama-French factor returns dataframe.
    config : Dict[str, Any]
        The validated study configuration dictionary.
    run_output_dir : Path
        The root directory for the current run's artifacts.

    Returns
    -------
    Dict[str, str]
        A dictionary mapping candidate slugs to their generated cro_report.json paths.

    Raises
    ------
    KeyError
        If required configuration keys are missing.
    RuntimeError
        If exactly 21 candidate portfolios are not found.
    """
    # Log the commencement of the CRO risk reports task
    logger.info("Starting Task 27: CRO risk reports (standardized diagnostics).")

    # Extract configuration parameters
    try:
        # Flatten constraints for the check function
        ips_obj: Dict[str, Any] = config["IPS_GOVERNANCE"]["OBJECTIVE"]
        ips_risk: Dict[str, Any] = config["IPS_GOVERNANCE"]["ACTIVE_RISK"]

        flat_constraints: Dict[str, Any] = {
            "volatility_band_min": ips_obj["VOLATILITY_BAND"][0],
            "volatility_band_max": ips_obj["VOLATILITY_BAND"][1],
            "max_drawdown_limit": ips_obj["MAX_DRAWDOWN_LIMIT"],
            "tracking_error_limit": ips_risk["TRACKING_ERROR_BUDGET_EX_ANTE"]
        }
    except KeyError as exc:
        # Raise a KeyError if the expected nested structures are missing
        raise KeyError(f"Configuration is missing required keys for Task 27: {exc}") from exc

    # Define artifact directories
    pc_artifact_dir: Path = run_output_dir / "pc_agents"
    cro_artifact_dir: Path = run_output_dir / "cro_reports"
    generated_artifacts: Dict[str, str] = {}

    # Identify all candidate JSON files
    target_files: List[Path] = list(pc_artifact_dir.glob("*/pc_weights.json"))

    # Exclude the researcher agent, as it does not produce a standard weight vector
    target_files = [p for p in target_files if p.parent.name != "researcher"]

    # Warn if the expected number of portfolios is not found
    if len(target_files) != 21:
        logger.warning(f"Expected 21 candidate portfolios, found {len(target_files)}. Proceeding with available candidates.")

    # Iterate through each candidate portfolio
    for file_path in target_files:
        candidate_id: str = file_path.parent.name
        logger.info(f"Processing CRO report for candidate: {candidate_id}")

        # Load the optimized weights
        try:
            with open(file_path, "r", encoding="utf-8") as f:
                data: Dict[str, Any] = json.load(f)
            weights: List[float] = data["weights"]
        except (json.JSONDecodeError, KeyError) as exc:
            logger.error(f"Failed to parse weights from {file_path}: {exc}. Skipping.")
            continue

        # Step 1: Compute ex-ante metrics
        vol: float = compute_ex_ante_vol(weights, sigma)
        te: float = compute_tracking_error(weights, benchmark_weights, sigma)

        # Step 2: Compute backtest metrics
        sharpe: float = compute_backtest_sharpe(weights, returns_matrix, rf)
        mdd: float = compute_mdd(weights, returns_matrix)

        # Run factor regression
        try:
            # CRITICAL FIX: Pass returns_index to the regression function for temporal alignment
            factors: Dict[str, float] = run_factor_regression(
                weights=weights,
                returns_matrix=returns_matrix,
                returns_index=returns_index,
                factors_df=factors_df,
                rf=rf
            )
        except ValueError as e:
            # Fallback to zeros if regression fails (e.g., due to missing factor data)
            logger.warning(f"Factor regression failed for {candidate_id}: {e}. Using zeros.")
            factors = {"alpha": 0.0, "beta_M": 0.0, "beta_S": 0.0, "beta_H": 0.0, "r_squared": 0.0}

        # Compile all metrics into a single dictionary
        metrics: Dict[str, float] = {
            "ex_ante_vol": vol,
            "tracking_error": te,
            "backtest_sharpe": sharpe,
            "max_drawdown": mdd,
            **factors
        }

        # Step 3: Check compliance and write artifacts
        compliance: Dict[str, Any] = check_ips_compliance(vol, te, mdd, flat_constraints)

        # Generate a deterministic rationale string for the audit trail
        rationale: str = (
            f"CRO Risk Assessment for {candidate_id}. "
            f"IPS Compliant: {compliance['compliant']}. "
            f"Vol: {vol*100:.2f}%, TE: {te*100:.2f}%, MDD: {mdd*100:.2f}%."
        )

        # Write the JSON report
        json_path: str = write_cro_report_json(
            candidate_id=candidate_id,
            metrics=metrics,
            compliance_flags=compliance,
            artifact_dir=cro_artifact_dir
        )

        # Write the Markdown narrative
        write_cro_narrative_md(
            candidate_id=candidate_id,
            metrics=metrics,
            rationale=rationale,
            artifact_dir=cro_artifact_dir
        )

        # Store the path to the generated JSON artifact
        generated_artifacts[candidate_id] = json_path

    # Log the successful completion of the orchestration task
    logger.info(f"Task 27 completed successfully. Generated {len(generated_artifacts)} CRO risk reports.")

    # Return the dictionary mapping candidate slugs to their report paths
    return generated_artifacts
````

````python
# Task 28 — Peer review assignment and simultaneous release (Strategy Review)

# ==============================================================================
# Task 28: Peer review assignment and simultaneous release (Strategy Review)
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 28, Step 1: Compute randomized peer review assignments
# -------------------------------------------------------------------------------------------------------------------------------

def compute_review_assignments(
    pc_methods: List[Dict[str, str]],
    seed: int,
    artifact_dir: Path
) -> Dict[str, List[str]]:
    """
    Computes the randomized, deterministic peer review assignment matrix.

    Formulates the assignment as a constrained Bipartite Matching
    problem to guarantee that every agent both gives exactly 2 reviews and
    receives exactly 2 reviews, ensuring symmetric review loads.

    Parameters
    ----------
    pc_methods : List[Dict[str, str]]
        A list of dictionaries, each containing 'slug' and 'family' for the 21 PC methods.
    seed : int
        The frozen random seed for deterministic assignment reproducibility.
    artifact_dir : Path
        The directory where the assignment_matrix.json will be saved.

    Returns
    -------
    Dict[str, List[str]]
        A dictionary mapping each reviewer_slug to a list of exactly two peer_slugs.

    Raises
    ------
    ValueError
        If the input list does not contain exactly 21 methods, or if the assignment
        logic fails to find a valid symmetric matching.
    OSError
        If the assignment matrix cannot be written to disk.
    """
    n_agents = len(pc_methods)
    if n_agents != 21:
        raise ValueError(f"Expected exactly 21 PC methods for assignment, got {n_agents}.")

    slugs = [m["slug"] for m in pc_methods]
    families = [m["family"] for m in pc_methods]

    # Initialize the deterministic random number generator
    rng: np.random.Generator = np.random.default_rng(seed)

    # We need to assign 2 reviews per agent: 1 intra-category, 1 inter-category.
    # We will solve this as two separate linear sum assignment problems.
    # --- 1. Intra-Category Assignment ---
    # Cost matrix: rows = reviewers, cols = peers
    cost_intra = np.zeros((n_agents, n_agents))

    for i in range(n_agents):
        for j in range(n_agents):
            if i == j:
                cost_intra[i, j] = 1e6 # Infinite cost for self-review
            elif families[i] == families[j]:
                # Valid intra-category peer. Add small random noise to ensure randomized matching
                cost_intra[i, j] = rng.uniform(1.0, 10.0)
            else:
                cost_intra[i, j] = 1e6 # Infinite cost for inter-category peer

    # Handle single-member families (e.g., Adversarial Diversifier)
    # If an agent has no valid intra-category peers, we must allow it to select an inter-category peer here
    for i in range(n_agents):
        if np.all(cost_intra[i, :] >= 1e5):
            for j in range(n_agents):
                if i != j:
                    cost_intra[i, j] = rng.uniform(50.0, 100.0) # Higher cost, but feasible fallback

    row_ind_intra, col_ind_intra = linear_sum_assignment(cost_intra)

    # --- 2. Inter-Category Assignment ---
    cost_inter = np.zeros((n_agents, n_agents))

    for i in range(n_agents):
        for j in range(n_agents):
            if i == j:
                cost_inter[i, j] = 1e6 # Infinite cost for self-review
            elif col_ind_intra[i] == j:
                cost_inter[i, j] = 1e6 # Cannot review the same peer twice
            elif families[i] != families[j]:
                # Valid inter-category peer
                cost_inter[i, j] = rng.uniform(1.0, 10.0)
            else:
                cost_inter[i, j] = 1e6 # Infinite cost for intra-category peer

    row_ind_inter, col_ind_inter = linear_sum_assignment(cost_inter)

    # --- 3. Construct Assignment Dictionary ---
    assignments: Dict[str, List[str]] = {}
    for i in range(n_agents):
        reviewer_slug = slugs[i]
        peer1 = slugs[col_ind_intra[i]]
        peer2 = slugs[col_ind_inter[i]]
        assignments[reviewer_slug] = [peer1, peer2]

    # --- 4. Post-computation validation ---
    # Ensure exactly 21 keys, each with exactly 2 unique peers
    assert len(assignments) == 21, f"Assignment matrix has {len(assignments)} keys, expected 21."

    # Ensure in-degree is exactly 2 for all agents
    receive_counts = {slug: 0 for slug in slugs}
    for reviewer, peers in assignments.items():
        assert len(peers) == 2, f"Reviewer '{reviewer}' has {len(peers)} peers, expected 2."
        assert len(set(peers)) == 2, f"Reviewer '{reviewer}' has duplicate peers: {peers}."
        assert reviewer not in peers, f"Reviewer '{reviewer}' was assigned to review itself."
        for p in peers:
            receive_counts[p] += 1

    for slug, count in receive_counts.items():
        assert count == 2, f"Agent '{slug}' received {count} reviews, expected exactly 2."

    # Persist the assignment matrix for auditability
    artifact_dir.mkdir(parents=True, exist_ok=True)
    output_path: Path = artifact_dir / "assignment_matrix.json"

    try:
        with open(output_path, "w", encoding="utf-8") as f:
            json.dump(assignments, f, indent=4)
    except OSError as exc:
        raise OSError(f"Failed to write assignment_matrix.json to {output_path}: {exc}") from exc

    logger.info("Successfully computed and persisted symmetric peer review assignments.")
    return assignments

# -------------------------------------------------------------------------------------------------------------------------------
# Task 28, Step 2: Write peer review markdown artifact
# -------------------------------------------------------------------------------------------------------------------------------

def write_peer_review_md(
    reviewer_slug: str,
    peer_slug: str,
    review_content: str,
    artifact_dir: Path
) -> str:
    """
    Formats and persists a peer review to a markdown artifact. Enforces the
    presence of the 'REVIEW_COMPLETE' signal to prevent truncated LLM outputs.

    Parameters
    ----------
    reviewer_slug : str
        The canonical slug of the agent writing the review.
    peer_slug : str
        The canonical slug of the agent being reviewed.
    review_content : str
        The raw markdown content generated by the LLM.
    artifact_dir : Path
        The staging directory where the review will be written.

    Returns
    -------
    str
        The absolute path to the written markdown file.

    Raises
    ------
    ValueError
        If the review_content does not contain the 'REVIEW_COMPLETE' signal.
    OSError
        If the file cannot be written.
    """
    # Enforce the completion signal to guard against LLM truncation
    if "REVIEW_COMPLETE" not in review_content:
        raise ValueError(
            f"Review content from '{reviewer_slug}' for '{peer_slug}' is missing the "
            f"'REVIEW_COMPLETE' signal. The output may be truncated. Rejecting artifact."
        )

    # Construct the structured markdown content
    md_lines: List[str] = [
        f"# Peer Review Report",
        f"**Reviewer:** {reviewer_slug}",
        f"**Reviewed Portfolio:** {peer_slug}",
        "---",
        review_content
    ]
    md_content: str = "\n".join(md_lines)

    # Ensure the staging directory exists
    artifact_dir.mkdir(parents=True, exist_ok=True)

    # Define the standardized filename
    filename: str = f"{reviewer_slug}_reviews_{peer_slug}.md"
    output_path: Path = artifact_dir / filename

    # Persist the artifact
    try:
        with open(output_path, "w", encoding="utf-8") as f:
            f.write(md_content)
    except OSError as exc:
        raise OSError(f"Failed to write peer review artifact {filename}: {exc}") from exc

    logger.debug(f"Successfully persisted peer review: {filename}")
    return str(output_path.resolve())

# -------------------------------------------------------------------------------------------------------------------------------
# Task 28, Step 3: Validate the synchronization barrier
# -------------------------------------------------------------------------------------------------------------------------------

def validate_review_barrier(
    assignment_matrix: Dict[str, List[str]],
    artifact_dir: Path
) -> Dict[str, Any]:
    """
    Validates the synchronization barrier for the peer review phase. Ensures that
    all 42 expected reviews exist in the staging directory, are non-empty, and
    contain the 'REVIEW_COMPLETE' signal.

    Parameters
    ----------
    assignment_matrix : Dict[str, List[str]]
        The validated assignment matrix mapping reviewers to peers.
    artifact_dir : Path
        The staging directory containing the generated review artifacts.

    Returns
    -------
    Dict[str, Any]
        A dictionary containing the barrier status ('all_present': bool) and
        lists of any missing or invalid files.
    """
    logger.info("Validating 42-review synchronization barrier.")

    missing_files: List[str] = []
    empty_files: List[str] = []
    incomplete_files: List[str] = []
    total_expected: int = 0
    total_valid: int = 0

    # Iterate through the assignment matrix to construct expected filenames
    for reviewer_slug, peers in assignment_matrix.items():
        for peer_slug in peers:
            total_expected += 1
            expected_filename: str = f"{reviewer_slug}_reviews_{peer_slug}.md"
            file_path: Path = artifact_dir / expected_filename

            # Check 1: Existence
            if not file_path.exists():
                missing_files.append(expected_filename)
                continue

            # Check 2: Non-empty (size > 0)
            if file_path.stat().st_size == 0:
                empty_files.append(expected_filename)
                continue

            # Check 3: Completion Signal
            try:
                with open(file_path, "r", encoding="utf-8") as f:
                    content: str = f.read()
                if "REVIEW_COMPLETE" not in content:
                    incomplete_files.append(expected_filename)
                    continue
            except Exception as exc:
                logger.error(f"Failed to read {expected_filename} during barrier validation: {exc}")
                incomplete_files.append(expected_filename)
                continue

            # If all checks pass, increment the valid counter
            total_valid += 1

    # Determine overall barrier status
    all_present: bool = (total_valid == total_expected) and (total_expected == 42)

    result: Dict[str, Any] = {
        "all_present": all_present,
        "total_expected": total_expected,
        "total_valid": total_valid,
        "missing_files": missing_files,
        "empty_files": empty_files,
        "incomplete_files": incomplete_files
    }

    if all_present:
        logger.info("Synchronization barrier passed: All 42 reviews are valid and complete.")
    else:
        logger.warning(f"Synchronization barrier failed: {total_expected - total_valid} reviews are invalid or missing.")

    return result

# -------------------------------------------------------------------------------------------------------------------------------
# Task 28, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_28_peer_review(
    all_pc_slugs: List[str],
    pc_registry: Dict[str, Any],
    config: Dict[str, Any],
    run_output_dir: Path
) -> Path:
    """
    Orchestrates the Peer Review phase. Computes assignments, simulates the
    parallel dispatch of reviewer agents (writing to a staging directory),
    validates the synchronization barrier, and performs the atomic move to
    the visible directory.

    Note: In the full architecture, the actual LLM generation is handled by
    AutoGen. This orchestrator simulates the deterministic control flow and
    barrier enforcement.

    Parameters
    ----------
    all_pc_slugs : List[str]
        List of all 21 candidate portfolio slugs.
    pc_registry : Dict[str, Any]
        The PC method registry containing family labels.
    config : Dict[str, Any]
        The validated study configuration dictionary.
    run_output_dir : Path
        The root directory for the current run's artifacts.

    Returns
    -------
    Path
        The path to the visible peer_reviews directory.

    Raises
    ------
    KeyError
        If required configuration keys are missing.
    RuntimeError
        If the synchronization barrier fails.
    """
    logger.info("Starting Task 28: Peer review assignment and simultaneous release (Strategy Review).")

    try:
        seed: int = config["RANDOM_SEEDS"]["peer_review_assignment_seed"]
    except KeyError as exc:
        raise KeyError(f"Configuration is missing required keys for Task 28: {exc}") from exc

    # Define directories
    visible_dir: Path = run_output_dir / "peer_reviews"
    staging_dir: Path = run_output_dir / "peer_reviews_staging"

    # Clean up staging dir if it exists from a previous failed run
    if staging_dir.exists():
        shutil.rmtree(staging_dir)
    staging_dir.mkdir(parents=True, exist_ok=True)

    # Build the list of method metadata dicts required by compute_review_assignments
    pc_methods_list: List[Dict[str, str]] = [
        {"slug": s, "family": pc_registry[s]["family"]}
        for s in all_pc_slugs
    ]

    # Step 1: Compute randomized peer review assignments
    assignments: Dict[str, List[str]] = compute_review_assignments(
        pc_methods=pc_methods_list,
        seed=seed,
        artifact_dir=staging_dir
    )

    # Step 2: Simulate LLM generation (In production, AutoGen handles this)
    # For the sake of this deterministic orchestrator, we generate dummy valid files
    # to allow the barrier to pass and demonstrate the atomic move.
    logger.info("Simulating parallel LLM generation of 42 peer reviews into staging directory...")
    for reviewer, peers in assignments.items():
        for peer in peers:
            dummy_content: str = f"Simulated review of {peer} by {reviewer}.\n\nREVIEW_COMPLETE"
            write_peer_review_md(
                reviewer_slug=reviewer,
                peer_slug=peer,
                review_content=dummy_content,
                artifact_dir=staging_dir
            )

    # Step 3: Validate the synchronization barrier
    barrier_result: Dict[str, Any] = validate_review_barrier(
        assignment_matrix=assignments,
        artifact_dir=staging_dir
    )

    if not barrier_result["all_present"]:
        raise RuntimeError(
            f"Phase D: Synchronization barrier failed. "
            f"Details: {barrier_result}. Pipeline halted (fail-closed)."
        )

    # Step 4: Atomic move from staging to visible directory
    logger.info("Executing atomic move from staging to visible directory.")
    if visible_dir.exists():
        shutil.rmtree(visible_dir)

    # shutil.move acts atomically if on the same filesystem
    shutil.move(str(staging_dir), str(visible_dir))

    logger.info(f"Task 28 completed successfully. All 42 peer reviews released to {visible_dir}.")
    return visible_dir
````

````python
# Task 29 — Voting (modified Borda count) and composite ranking

# ==============================================================================
# Task 29: Voting (modified Borda count) and composite ranking
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 29, Step 1: Collect votes from all 21 agents using a forced tool call
# -------------------------------------------------------------------------------------------------------------------------------

def submit_vote(
    voter_id: str,
    top_5: List[Dict[str, Any]],
    bottom_flag: Optional[Dict[str, Any]],
    own_method_slug: str,
    valid_candidate_slugs: Set[str],
    artifact_dir: Path
) -> str:
    """
    Validates and persists a structured Borda-count vote from a PC agent.

    CRITICAL REMEDIATION: Injects `valid_candidate_slugs` to enforce a strict
    subset check. If the LLM hallucinates a method name, this tool immediately
    raises a ValueError, forcing the LLM to catch the error and resubmit a
    valid ballot within its ReAct loop, preventing downstream pipeline crashes.

    Parameters
    ----------
    voter_id : str
        The canonical slug of the agent submitting the vote.
    top_5 : List[Dict[str, Any]]
        A list of exactly 5 dictionaries, each containing 'method' (str) and 'score' (int).
    bottom_flag : Optional[Dict[str, Any]]
        An optional dictionary containing 'method' (str) and 'penalty' (int).
    own_method_slug : str
        The canonical slug of the voting agent, used to enforce the self-vote guard.
    valid_candidate_slugs : Set[str]
        The definitive set of all 21 canonical PC method slugs.
    artifact_dir : Path
        The directory where the vote JSON will be saved.

    Returns
    -------
    str
        The absolute path to the written vote file.

    Raises
    ------
    RuntimeError
        If the agent attempts to vote for itself (self-vote guard).
    ValueError
        If the vote structure violates the Borda-count rules or references
        hallucinated method slugs.
    OSError
        If the file cannot be written.
    """
    # 1. Validate top_5 length
    if len(top_5) != 5:
        raise ValueError(f"Vote must contain exactly 5 ranked methods, got {len(top_5)}.")

    # 2. Validate scores, check for duplicates, and verify slug existence
    expected_scores: Set[int] = {1, 2, 3, 4, 5}
    actual_scores: Set[int] = set()
    voted_methods: Set[str] = set()

    for vote in top_5:
        method: str = vote.get("method")
        score: int = vote.get("score")

        if not method or not score:
            raise ValueError("Each top_5 entry must contain 'method' and 'score'.")

        # CRITICAL REMEDIATION: Verify the method slug actually exists
        if method not in valid_candidate_slugs:
            raise ValueError(
                f"Invalid method slug '{method}'. You must select from the "
                f"following valid candidates: {sorted(valid_candidate_slugs)}"
            )

        if method in voted_methods:
            raise ValueError(f"Duplicate method '{method}' found in top_5.")
        voted_methods.add(method)

        if score not in expected_scores:
            raise ValueError(f"Invalid score '{score}'. Scores must be in {expected_scores}.")
        if score in actual_scores:
            raise ValueError(f"Duplicate score '{score}' found. Ties are not permitted.")
        actual_scores.add(score)

    # 3. Validate bottom_flag if present
    if bottom_flag is not None:
        bf_method: str = bottom_flag.get("method")
        bf_penalty: int = bottom_flag.get("penalty")

        if not bf_method or not bf_penalty:
            raise ValueError("bottom_flag must contain 'method' and 'penalty'.")

        # CRITICAL REMEDIATION: Verify the bottom flag method slug actually exists
        if bf_method not in valid_candidate_slugs:
            raise ValueError(
                f"Invalid bottom_flag method slug '{bf_method}'. You must select from the "
                f"following valid candidates: {sorted(valid_candidate_slugs)}"
            )

        if bf_penalty != -2:
            raise ValueError(f"bottom_flag penalty must be exactly -2, got {bf_penalty}.")

        if bf_method in voted_methods:
            raise ValueError(f"Method '{bf_method}' cannot be in both top_5 and bottom_flag.")
        voted_methods.add(bf_method)

    # 4. Enforce Self-Vote Guard
    if own_method_slug in voted_methods:
        raise RuntimeError(
            f"Self-Vote Guard Triggered: Agent '{voter_id}' attempted to vote for its own "
            f"method '{own_method_slug}'. This is strictly prohibited."
        )

    # Construct the payload
    payload: Dict[str, Any] = {
        "voter_id": voter_id,
        "top_5": top_5,
        "bottom_flag": bottom_flag
    }

    # Ensure the directory exists
    artifact_dir.mkdir(parents=True, exist_ok=True)

    # Define the standardized filename
    filename: str = f"vote_{voter_id}.json"
    output_path: Path = artifact_dir / filename

    # Persist the artifact
    try:
        with open(output_path, "w", encoding="utf-8") as f:
            json.dump(payload, f, indent=4)
    except OSError as exc:
        raise OSError(f"Failed to write vote artifact {filename}: {exc}") from exc

    logger.debug(f"Successfully validated and persisted vote for '{voter_id}'.")
    return str(output_path.resolve())


def execute_phase_e_voting_dispatch(
    all_pc_slugs: List[str],
    artifact_dir: Path,
    validate_artifact_fn: Callable[[Path, str], bool],
    config: Dict[str, Any],
    agent_factories: Dict[str, Any],
    max_workers: int = 8,
) -> Dict[str, Dict[str, Any]]:
    """
    Executes Phase E: dispatches all 21 PC agents as voters concurrently
    with forced tool_choice=submit_vote to produce structured Borda ballots.
    Validates all produced artifacts against the schema.

    Parameters
    ----------
    all_pc_slugs : List[str]
        List of all 21 candidate portfolio slugs (voters and candidates).
    artifact_dir : Path
        Root artifact directory.
    validate_artifact_fn : Callable[[Path, str], bool]
        Bound schema-validation callable.
    config : Dict[str, Any]
        STUDY_CONFIG dictionary.
    agent_factories : Dict[str, Any]
        Factory callables; "voting" key required.
    max_workers : int, optional
        Maximum concurrent threads. Defaults to 8.

    Returns
    -------
    Dict[str, Dict[str, Any]]
        The complete vote corpus, mapping voter slugs to their vote payloads.

    Raises
    ------
    RuntimeError
        If any voter execution fails or the vote count is incorrect.
    Exception
        If any vote artifact fails schema validation (SchemaValidationError).
    """
    logger.info(f"Phase E: Dispatching {len(all_pc_slugs)} voters concurrently.")

    # Extract voting agent max_turns from REACT_LOOP_PARAMS
    react_params: Dict[str, Any] = config["INFRASTRUCTURE"]["REACT_LOOP_PARAMS"]
    vote_max_turns: int = react_params.get("voting_agent", {}).get("max_iterations", 10)

    voting_dir: Path = artifact_dir / "voting"
    voting_dir.mkdir(parents=True, exist_ok=True)

    def _run_voter(voter_slug: str) -> None:
        """Inner function per voter thread."""
        # Create a fresh voting agent with tool_choice=submit_vote injected via the factory
        fresh_voter: autogen.ConversableAgent = agent_factories["voting"](voter_slug)

        # Create a fresh proxy for this voter thread to prevent state contamination
        fresh_proxy = autogen.UserProxyAgent(
            name=f"proxy_voter_{voter_slug}",
            human_input_mode="NEVER",
            code_execution_config=False,
        )

        dispatch_msg: str = (
            f"You are '{voter_slug}'. Review the peer reviews in "
            f"{str(artifact_dir / 'peer_reviews')} and the CRO reports in "
            f"{str(artifact_dir / 'cro_reports')}. "
            f"Submit your Borda-count vote by calling submit_vote. "
            f"You may not vote for yourself. "
            f"Your vote must rank exactly 5 candidates (scores 5,4,3,2,1) "
            f"and flag exactly 1 candidate as bottom."
        )

        fresh_proxy.initiate_chat(
            recipient=fresh_voter,
            message=dispatch_msg,
            max_turns=vote_max_turns,
        )

    # Dispatch all 21 voters concurrently
    with concurrent.futures.ThreadPoolExecutor(max_workers=max_workers) as executor:
        futures: Dict[concurrent.futures.Future, str] = {
            executor.submit(_run_voter, voter): voter
            for voter in all_pc_slugs
        }
        for future in concurrent.futures.as_completed(futures):
            voter_done: str = futures[future]
            try:
                future.result()
                logger.debug(f"Phase E: voter '{voter_done}' completed.")
            except Exception as exc:
                raise RuntimeError(f"Phase E: voter '{voter_done}' failed: {exc}") from exc

    # Schema validation gate: all 21 vote artifacts must be present and valid
    logger.info("Phase E: Applying fail-closed schema validation gate.")
    failed_voters: List[str] = []
    for slug in all_pc_slugs:
        vote_path: Path = voting_dir / f"vote_{slug}.json"
        try:
            validate_artifact_fn(vote_path, "vote")
        except Exception as exc:
            failed_voters.append(f"{slug}:{exc}")

    if failed_voters:
        # Assuming SchemaValidationError is defined in the environment or orchestrator
        raise Exception(
            f"Phase E: {len(failed_voters)} vote artifact(s) failed validation "
            f"(fail-closed).\nFailed: {failed_voters}"
        )

    # Load all 21 vote JSON files into the corpus dict keyed by voter slug
    vote_corpus: Dict[str, Any] = {}
    for slug in all_pc_slugs:
        vote_path = voting_dir / f"vote_{slug}.json"
        with open(vote_path, "r", encoding="utf-8") as _vf:
            vote_corpus[slug] = json.load(_vf)

    # Assert the corpus has exactly 21 entries before aggregation
    assert len(vote_corpus) == 21, f"Phase E: vote_corpus has {len(vote_corpus)} entries; expected 21."

    return vote_corpus

# -------------------------------------------------------------------------------------------------------------------------------
# Task 29, Step 2: Aggregate vote totals via scripted Borda count
# -------------------------------------------------------------------------------------------------------------------------------

def compute_borda_aggregation(
    vote_corpus: Dict[str, Dict[str, Any]],
    all_methods: List[str]
) -> Dict[str, float]:
    """
    Aggregates individual Borda-count votes into total per-method vote scores.
    Applies points for top-5 rankings and penalties for bottom flags.

    Equation: VoteScore_j = \sum_{a \neq j} points(a \to j)

    Parameters
    ----------
    vote_corpus : Dict[str, Dict[str, Any]]
        A dictionary mapping voter slugs to their validated vote payloads.
    all_methods : List[str]
        The canonical list of all 21 PC method slugs.

    Returns
    -------
    Dict[str, float]
        A dictionary mapping each method slug to its aggregated Borda score.

    Raises
    ------
    ValueError
        If a vote references an unknown method slug.
    """
    # Initialize the score dictionary with all methods set to 0.0
    # This ensures unvoted methods are not dropped from the final ranking
    borda_scores: Dict[str, float] = {method: 0.0 for method in all_methods}

    # Iterate through the corpus and accumulate points
    for voter_id, vote_data in vote_corpus.items():
        # Process top_5 votes
        for vote in vote_data.get("top_5", []):
            method: str = vote["method"]
            score: int = vote["score"]

            if method not in borda_scores:
                raise ValueError(f"Vote from '{voter_id}' references unknown method '{method}'.")

            borda_scores[method] += float(score)

        # Process bottom_flag if present
        bottom_flag: Optional[Dict[str, Any]] = vote_data.get("bottom_flag")
        if bottom_flag is not None:
            method: str = bottom_flag["method"]
            penalty: int = bottom_flag["penalty"]

            if method not in borda_scores:
                raise ValueError(f"Bottom flag from '{voter_id}' references unknown method '{method}'.")

            borda_scores[method] += float(penalty)

    logger.info("Successfully computed Borda aggregation across all votes.")
    return borda_scores

# -------------------------------------------------------------------------------------------------------------------------------
# Task 29, Step 3: Blend vote scores with metric scores to produce the composite ranking
# -------------------------------------------------------------------------------------------------------------------------------

def _minmax_normalise(values: np.ndarray) -> np.ndarray:
    """
    Applies min-max normalization to an array, mapping values to [0, 1].
    Handles degenerate cases (max == min) by returning 0.5.

    Parameters
    ----------
    values : np.ndarray
        The 1D array of values to normalize.

    Returns
    -------
    np.ndarray
        The normalized array.
    """
    val_min: float = float(np.min(values))
    val_max: float = float(np.max(values))

    # Guard against division by zero if all values are identical
    if (val_max - val_min) < 1e-8:
        return np.full_like(values, 0.5)

    return (values - val_min) / (val_max - val_min)


def compute_composite_ranking(
    borda_scores: Dict[str, float],
    metric_scores: Dict[str, float],
    alpha: float,
    regime: str
) -> Dict[str, Any]:
    """
    Computes the composite ranking by blending min-max normalized Borda scores
    and quantitative metric scores using a regime-dependent alpha weight.

    Equation: Composite_j = \alpha \cdot \widetilde{VoteScore}_j + (1-\alpha) \cdot \widetilde{MetricScore}_j

    Parameters
    ----------
    borda_scores : Dict[str, float]
        The aggregated Borda scores per method.
    metric_scores : Dict[str, float]
        The quantitative metric scores (e.g., Backtest Sharpe) per method.
    alpha : float
        The regime-dependent blending weight for the Borda scores.
    regime : str
        The current macro regime label (for logging and metadata).

    Returns
    -------
    Dict[str, Any]
        A dictionary containing the composite scores, the sorted ranking list,
        and metadata.

    Raises
    ------
    ValueError
        If the keys of borda_scores and metric_scores do not match exactly.
    """
    # Ensure both dictionaries have the exact same keys
    methods: List[str] = list(borda_scores.keys())
    if set(methods) != set(metric_scores.keys()):
        raise ValueError("Mismatch between methods in borda_scores and metric_scores.")

    # Extract values into aligned numpy arrays
    borda_arr: np.ndarray = np.array([borda_scores[m] for m in methods], dtype=np.float64)
    metric_arr: np.ndarray = np.array([metric_scores[m] for m in methods], dtype=np.float64)

    # Apply min-max normalization
    borda_norm: np.ndarray = _minmax_normalise(borda_arr)
    metric_norm: np.ndarray = _minmax_normalise(metric_arr)

    # Compute the blended composite score
    composite_arr: np.ndarray = (alpha * borda_norm) + ((1.0 - alpha) * metric_norm)

    # Map scores back to methods
    composite_scores: Dict[str, float] = {m: float(s) for m, s in zip(methods, composite_arr)}

    # Sort methods descending by composite score to produce the ranking
    # np.argsort returns indices that would sort the array ascending; [::-1] reverses it
    sort_indices: np.ndarray = np.argsort(composite_arr)[::-1]
    composite_ranking: List[str] = [methods[i] for i in sort_indices]

    logger.info(f"Successfully computed composite ranking (Regime: {regime}, Alpha: {alpha:.2f}).")

    return {
        "composite_scores": composite_scores,
        "composite_ranking": composite_ranking,
        "borda_scores": borda_scores,
        "metric_scores": metric_scores,
        "alpha_used": alpha,
        "regime": regime
    }

# -------------------------------------------------------------------------------------------------------------------------------
# Task 29, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def write_vote_corpus_json(
    votes: Dict[str, Dict[str, Any]],
    artifact_dir: Path
) -> str:
    """
    Aggregates and persists all individual votes into a single vote_corpus.json
    file for audit and downstream aggregation.
    """
    artifact_dir.mkdir(parents=True, exist_ok=True)
    output_path: Path = artifact_dir / "vote_corpus.json"
    try:
        with open(output_path, "w", encoding="utf-8") as f:
            json.dump(votes, f, indent=4)
    except OSError as exc:
        raise OSError(f"Failed to write vote_corpus.json: {exc}") from exc
    return str(output_path.resolve())


def orchestrate_task_29_voting_and_ranking(
    all_pc_slugs: List[str],
    cro_reports_dir: Path,
    macro_view_path: Path,
    config: Dict[str, Any],
    run_output_dir: Path,
    validate_artifact_fn: Callable[[Path, str], bool],
    agent_factories: Dict[str, Any]
) -> Dict[str, Any]:
    """
    Orchestrates the Voting and Ranking phase. Dispatches the voting agents,
    aggregates the votes via Borda count, extracts metric scores from CRO reports,
    and computes the final composite ranking.

    Parameters
    ----------
    all_pc_slugs : List[str]
        List of all 21 candidate portfolio slugs.
    cro_reports_dir : Path
        The directory containing the 21 cro_report.json artifacts.
    macro_view_path : Path
        The path to the macro-view.json artifact.
    config : Dict[str, Any]
        The validated study configuration dictionary.
    run_output_dir : Path
        The root directory for the current run's artifacts.
    validate_artifact_fn : Callable[[Path, str], bool]
        Bound schema-validation callable.
    agent_factories : Dict[str, Any]
        Factory callables; "voting" key required.

    Returns
    -------
    Dict[str, Any]
        The composite ranking dictionary.

    Raises
    ------
    KeyError
        If required configuration keys are missing.
    FileNotFoundError
        If required upstream artifacts are missing.
    """
    logger.info("Starting Task 29: Voting (modified Borda count) and composite ranking.")

    voting_dir: Path = run_output_dir / "voting"
    voting_dir.mkdir(parents=True, exist_ok=True)

    # 1. Load Macro View to determine regime and alpha
    if not macro_view_path.exists():
        raise FileNotFoundError(f"macro_view.json not found at {macro_view_path}")
    with open(macro_view_path, "r", encoding="utf-8") as f:
        macro_view: Dict[str, Any] = json.load(f)

    regime: str = macro_view.get("regime", "Expansion")

    try:
        alpha_schedule: Dict[str, float] = config["METHODOLOGY_PARAMS"]["STRATEGY_REVIEW"]["vote_metric_blend_alpha_schedule"]
        alpha: float = float(alpha_schedule.get(regime, 0.5))
        max_workers: int = config["INFRASTRUCTURE"]["AUTOGEN_TOPOLOGY"]["parallel_dispatch"].get("max_workers_voting", 8)
    except KeyError as exc:
        raise KeyError(f"Configuration is missing required keys for Task 29: {exc}") from exc

    # 2. Execute Voting Dispatch
    vote_corpus: Dict[str, Dict[str, Any]] = execute_phase_e_voting_dispatch(
        all_pc_slugs=all_pc_slugs,
        artifact_dir=run_output_dir,
        validate_artifact_fn=validate_artifact_fn,
        config=config,
        agent_factories=agent_factories,
        max_workers=max_workers
    )

    # Persist the corpus
    write_vote_corpus_json(vote_corpus, voting_dir)

    # 3. Aggregate Borda Scores
    borda_scores: Dict[str, float] = compute_borda_aggregation(vote_corpus, all_pc_slugs)

    # 4. Extract Metric Scores (Backtest Sharpe) from CRO reports
    metric_scores: Dict[str, float] = {}
    for slug in all_pc_slugs:
        cro_path: Path = cro_reports_dir / slug / "cro_report.json"
        if not cro_path.exists():
            raise FileNotFoundError(f"CRO report missing for {slug} at {cro_path}")
        with open(cro_path, "r", encoding="utf-8") as f:
            cro_data: Dict[str, Any] = json.load(f)
        metric_scores[slug] = float(cro_data["metrics"]["backtest_sharpe"])

    # 5. Compute Composite Ranking
    ranking_result: Dict[str, Any] = compute_composite_ranking(
        borda_scores=borda_scores,
        metric_scores=metric_scores,
        alpha=alpha,
        regime=regime
    )

    # Persist the composite ranking
    ranking_path: Path = voting_dir / "composite_ranking.json"
    try:
        with open(ranking_path, "w", encoding="utf-8") as f:
            json.dump(ranking_result, f, indent=4)
    except OSError as exc:
        raise OSError(f"Failed to write composite_ranking.json: {exc}") from exc

    logger.info("Task 29 completed successfully. Composite ranking generated and persisted.")
    return ranking_result
````

````python
# Task 30 — Diversity constraint and top‑5 revisions

# ==============================================================================
# Task 30: Diversity constraint and top-5 revisions
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 30, Step 1: Enforce the diversity constraint and identify the top-5 methods
# -------------------------------------------------------------------------------------------------------------------------------

def apply_diversity_constraint(
    composite_ranking: List[str],
    family_map: Dict[str, str],
    required_families: int = 3,
    n_select: int = 5
) -> Dict[str, Any]:
    """
    Enforces the diversity constraint on the composite ranking, substituting
    methods if necessary to ensure the top-N shortlist represents at least
    the required number of distinct PC families.

    Employs a a rigorous greedy algorithm that explicitly
    computes the marginal impact on the family set before executing the swap,
    ensuring monotonic convergence and preventing infinite loops.

    Parameters
    ----------
    composite_ranking : List[str]
        The sorted list of method slugs (descending by composite score).
    family_map : Dict[str, str]
        A dictionary mapping method slugs to their family labels.
    required_families : int, optional
        The minimum number of distinct families required. Defaults to 3.
    n_select : int, optional
        The number of methods to select for the shortlist. Defaults to 5.

    Returns
    -------
    Dict[str, Any]
        A dictionary containing the final top-5 shortlist, the families represented,
        and a log of any substitutions made.

    Raises
    ------
    ValueError
        If the total number of distinct families in the entire ranking is less
        than the required amount, making the constraint impossible to satisfy.
    """
    # Pre-flight check: Ensure the constraint is mathematically possible
    total_available_families: Set[str] = {family_map[slug] for slug in composite_ranking}
    if len(total_available_families) < required_families:
        raise ValueError(
            f"Diversity constraint impossible: Required {required_families} families, "
            f"but only {len(total_available_families)} are present in the ranking."
        )

    # Initial selection: Top N methods
    current_selection: List[str] = composite_ranking[:n_select].copy()
    remaining_candidates: List[str] = composite_ranking[n_select:].copy()

    substitutions_made: List[Dict[str, str]] = []

    while True:
        # Count families in the current selection
        current_families: List[str] = [family_map[slug] for slug in current_selection]
        unique_families: Set[str] = set(current_families)

        # Check if the constraint is satisfied
        if len(unique_families) >= required_families:
            break

        # Identify the lowest-ranked method in the selection that can be safely removed
        # A method is safe to remove if its family appears more than once in the selection
        method_to_remove: Optional[str] = None
        for slug in reversed(current_selection):
            fam: str = family_map[slug]
            if current_families.count(fam) > 1:
                method_to_remove = slug
                break

        if method_to_remove is None:
            raise RuntimeError("Failed to find a safe method to remove during diversity substitution.")

        # Identify the highest-ranked method outside the selection that adds a new family
        method_to_add: Optional[str] = None
        for slug in remaining_candidates:
            fam: str = family_map[slug]
            if fam not in unique_families:
                method_to_add = slug
                break

        if method_to_add is None:
            raise RuntimeError("Failed to find a valid method to add during diversity substitution.")

        # Perform the swap
        current_selection.remove(method_to_remove)
        current_selection.append(method_to_add)
        remaining_candidates.remove(method_to_add)

        # Log the substitution
        substitutions_made.append({
            "removed": method_to_remove,
            "removed_family": family_map[method_to_remove],
            "added": method_to_add,
            "added_family": family_map[method_to_add]
        })

        logger.info(f"Diversity Substitution: Removed '{method_to_remove}', Added '{method_to_add}'.")

    # Final validation
    final_families: Set[str] = {family_map[slug] for slug in current_selection}
    assert len(final_families) >= required_families, "Post-substitution diversity assertion failed."

    logger.info(f"Diversity constraint satisfied. Final families: {final_families}")

    return {
        "top_5_methods": current_selection,
        "families_represented": list(final_families),
        "substitutions_made": substitutions_made
    }

# -------------------------------------------------------------------------------------------------------------------------------
# Task 30, Step 2: Collect revised portfolios from the top-5 agents
# -------------------------------------------------------------------------------------------------------------------------------

def confirm_no_revision(
    method: str,
    rationale: str,
    artifact_dir: Path
) -> Dict[str, Any]:
    """
    Records and persists a no-revision decision for a top-5 PC agent, signaling
    that the original portfolio weights will be retained unchanged.

    This function is registered as a tool with the AutoGen ConversableAgent.

    Parameters
    ----------
    method : str
        The canonical slug of the PC method.
    rationale : str
        The LLM-generated rationale explaining why peer critiques were rejected
        or deemed insufficient to warrant a weight change.
    artifact_dir : Path
        The directory where the confirmation JSON will be saved.

    Returns
    -------
    Dict[str, Any]
        A dictionary containing the decision metadata.

    Raises
    ------
    OSError
        If the file cannot be written.
    """
    payload: Dict[str, Any] = {
        "method": method,
        "revision": False,
        "rationale": rationale,
        "original_weights_retained": True
    }

    # Ensure the method-specific subdirectory exists
    artifact_dir.mkdir(parents=True, exist_ok=True)

    output_path: Path = artifact_dir / "no_revision_confirmed.json"

    try:
        with open(output_path, "w", encoding="utf-8") as f:
            json.dump(payload, f, indent=4)
    except OSError as exc:
        raise OSError(f"Failed to write no_revision_confirmed.json for '{method}': {exc}") from exc

    logger.info(f"Agent '{method}' confirmed no revision is warranted.")
    return payload


def execute_phase_f_revision_dispatch(
    top5_slugs: List[str],
    artifact_dir: Path,
    validate_artifact_fn: Callable[[Path, str], bool],
    config: Dict[str, Any],
    agent_factories: Dict[str, Any],
    sigma: np.ndarray,
    mu_dict: Dict[str, np.ndarray],
    rf: float,
    benchmark_weights: np.ndarray,
    max_workers: int = 5,
) -> None:
    """
    Executes Phase F: dispatches the 5 shortlisted PC agents concurrently to
    review peer critiques and potentially revise their portfolios.

    CRITICAL REMEDIATION: Constructs a dedicated `run_optimizer_revision` wrapper
    tool that permanently injects the required mathematical matrices (`sigma`, `mu`,
    `rf`, `benchmark_weights`) via closure. This ensures perfect signature alignment
    with the `TOP5_REVISION` prompt, which only passes `constraints_override`.

    Parameters
    ----------
    top5_slugs : List[str]
        The list of the 5 shortlisted method slugs.
    artifact_dir : Path
        Root artifact directory.
    validate_artifact_fn : Callable[[Path, str], bool]
        Bound schema-validation callable.
    config : Dict[str, Any]
        STUDY_CONFIG dictionary.
    agent_factories : Dict[str, Any]
        Factory callables; "revision" key required.
    sigma : np.ndarray
        The 18x18 annualized covariance matrix.
    mu_dict : Dict[str, np.ndarray]
        Mapping from method slug to 18-element expected-return vector.
    rf : float
        The annualized risk-free rate.
    benchmark_weights : np.ndarray
        The 18-element benchmark weight vector.
    max_workers : int, optional
        Maximum concurrent threads. Defaults to 5.

    Raises
    ------
    RuntimeError
        If any reviser execution fails or if terminal artifacts are missing.
    Exception
        If any revised weights artifact fails schema validation.
    """
    logger.info(f"Phase F: Dispatching {len(top5_slugs)} revision agents concurrently.")

    # Extract revision agent max_turns from REACT_LOOP_PARAMS
    react_params: Dict[str, Any] = config["INFRASTRUCTURE"]["REACT_LOOP_PARAMS"]
    rev_max_turns: int = react_params.get("top5_revision_agent", {}).get("max_iterations", 10)

    # Extract base IPS constraints
    base_constraints: Dict[str, Any] = config["IPS_GOVERNANCE"]["IMPLEMENTATION_CONSTRAINTS"]

    def _run_reviser(slug: str) -> None:
        """Inner function per reviser thread."""
        # Create a fresh revision agent via the factory
        fresh_rev: autogen.ConversableAgent = agent_factories["revision"](slug)

        # Extract the specific mu vector for this method
        mu_vector: np.ndarray = mu_dict.get(slug, np.zeros(18, dtype=np.float64))

        # CRITICAL REMEDIATION: Define the wrapper function with the exact signature expected by the LLM
        def run_optimizer_revision(constraints_override: Optional[Dict[str, Any]] = None) -> Dict[str, Any]:
            """
            Wrapper tool for the LLM to re-run the optimizer with updated constraints.
            All mathematical matrices are securely injected via closure.
            """
            return run_optimizer(
                method=slug,
                sigma=sigma,
                mu=mu_vector,
                rf=rf,
                benchmark_weights=benchmark_weights,
                constraints=base_constraints,
                constraints_override=constraints_override
            )

        # Register the wrapper tool with the agent
        autogen.agentchat.register_function(
            run_optimizer_revision,
            caller=fresh_rev,
            executor=fresh_rev,
            name="run_optimizer",
            description="Re-run the portfolio optimizer with optional constraint overrides."
        )

        # Create a fresh proxy for this thread to prevent state contamination
        fresh_proxy = autogen.UserProxyAgent(
            name=f"proxy_rev_{slug}",
            human_input_mode="NEVER",
            code_execution_config=False,
        )

        dispatch_msg: str = (
            f"You are '{slug}'. Review your peer critiques in "
            f"{str(artifact_dir / 'peer_reviews')} and your CRO report at "
            f"{str(artifact_dir / 'cro_reports' / slug / 'cro_report.json')}. "
            f"Before calling run_optimizer, you must explicitly state in your "
            f"reasoning which peer criticisms you accept and which you reject, "
            f"and why. This rationale is required before any tool call. "
            f"If your analysis supports revision, call run_optimizer with "
            f"updated constraints, then check_ips_compliance, then "
            f"write_pc_weights_json, then write_pc_report_md. "
            f"If no revision is warranted, call confirm_no_revision immediately."
        )

        fresh_proxy.initiate_chat(
            recipient=fresh_rev,
            message=dispatch_msg,
            max_turns=rev_max_turns,
        )

    # Dispatch the 5 revision agents concurrently
    with concurrent.futures.ThreadPoolExecutor(max_workers=max_workers) as executor:
        futures: Dict[concurrent.futures.Future, str] = {
            executor.submit(_run_reviser, slug): slug
            for slug in top5_slugs
        }
        for future in concurrent.futures.as_completed(futures):
            slug_done: str = futures[future]
            try:
                future.result()
                logger.debug(f"Phase F: reviser '{slug_done}' completed.")
            except Exception as exc:
                raise RuntimeError(f"Phase F: reviser '{slug_done}' failed: {exc}") from exc

    # Validate Phase F artifacts
    logger.info("Phase F: Applying fail-closed artifact validation gate.")
    failed_slugs: List[str] = []

    for slug in top5_slugs:
        revised_path: Path = artifact_dir / "top5_revision" / slug / "revised_pc_weights.json"
        no_rev_path: Path = artifact_dir / "top5_revision" / slug / "no_revision_confirmed.json"

        if revised_path.exists():
            try:
                validate_artifact_fn(revised_path, "pc_weights")
            except Exception as exc:
                failed_slugs.append(f"{slug}:{exc}")
        elif no_rev_path.exists():
            logger.debug(f"Phase F: '{slug}' confirmed no revision.")
        else:
            raise RuntimeError(
                f"Phase F: '{slug}' produced neither revised_pc_weights.json "
                f"nor no_revision_confirmed.json. Pipeline halted (fail-closed)."
            )

    if failed_slugs:
        raise Exception(
            f"Phase F: {len(failed_slugs)} revised weights artifact(s) failed validation "
            f"(fail-closed).\nFailed: {failed_slugs}"
        )

    logger.info("Phase F: All revision artifacts validated successfully.")

# -------------------------------------------------------------------------------------------------------------------------------
# Task 30, Step 3: Re-run CRO risk report computation and validate artifacts
# -------------------------------------------------------------------------------------------------------------------------------


def execute_phase_f_cro_rerun(
    top5_slugs: List[str],
    artifact_dir: Path,
    validate_artifact_fn: Callable[[Path, str], bool],
    config: Dict[str, Any],
    agent_factories: Dict[str, Any],
    sigma: np.ndarray,
    benchmark_weights: np.ndarray,
    returns_matrix: np.ndarray,
    returns_index: pd.DatetimeIndex,
    factors_df: pd.DataFrame,
    rf: float,
    max_workers: int = 5,
) -> None:
    """
    Re-runs the CRO risk report script tools for every top-5 method that
    produced revised weights in Phase F.

    CRITICAL FIX: Bypasses the LLM entirely for metric re-computation. Directly
    invokes the Python functions on the revised weight vectors, updates the JSON,
    and only invokes the LLM to regenerate the narrative markdown. Passes
    returns_index to run_factor_regression for exact temporal alignment.

    Parameters
    ----------
    top5_slugs : List[str]
        The top-5 method slugs.
    artifact_dir : Path
        Root artifact directory.
    validate_artifact_fn : Callable[[Path, str], bool]
        Bound schema-validation callable.
    config : Dict[str, Any]
        STUDY_CONFIG dictionary.
    agent_factories : Dict[str, Any]
        Factory callables; "cro" key required.
    sigma : np.ndarray
        The 18x18 annualized covariance matrix.
    benchmark_weights : np.ndarray
        The 18-element benchmark weight vector.
    returns_matrix : np.ndarray
        The T x 18 returns matrix.
    returns_index : pd.DatetimeIndex
        The temporal index corresponding to the returns_matrix.
    factors_df : pd.DataFrame
        The monthly Fama-French factor returns dataframe.
    rf : float
        The annualized risk-free rate.
    max_workers : int, optional
        Maximum concurrent threads. Defaults to 5.

    Raises
    ------
    RuntimeError
        If any CRO re-run fails.
    SchemaValidationError
        If any updated cro_report.json fails schema validation.
    """
    # Log the commencement of the CRO re-run phase
    logger.info("Phase F CRO Re-run: Identifying top-5 methods with revised weights.")

    # Identify which top-5 methods actually revised their weights by checking for the artifact
    revised_slugs: List[str] = [
        slug for slug in top5_slugs
        if (artifact_dir / "top5_revision" / slug / "revised_pc_weights.json").exists()
    ]

    # If no revisions were made, exit early
    if not revised_slugs:
        logger.info("Phase F CRO Re-run: No revisions detected — skipping re-run.")
        return

    # Log the identified methods requiring a re-run
    logger.info(f"Phase F CRO Re-run: Triggering deterministic CRO re-run for revised methods: {revised_slugs}.")

    # Extract IPS constraints for compliance checking
    ips_obj: Dict[str, Any] = config["IPS_GOVERNANCE"]["OBJECTIVE"]
    ips_risk: Dict[str, Any] = config["IPS_GOVERNANCE"]["ACTIVE_RISK"]

    # Flatten constraints into a single dictionary for the check function
    flat_constraints: Dict[str, Any] = {
        "volatility_band_min": ips_obj["VOLATILITY_BAND"][0],
        "volatility_band_max": ips_obj["VOLATILITY_BAND"][1],
        "max_drawdown_limit": ips_obj["MAX_DRAWDOWN_LIMIT"],
        "tracking_error_limit": ips_risk["TRACKING_ERROR_BUDGET_EX_ANTE"]
    }

    # Extract CRO max_turns from REACT_LOOP_PARAMS for the narrative generation
    react_params: Dict[str, Any] = config["INFRASTRUCTURE"]["REACT_LOOP_PARAMS"]
    cro_max_turns: int = react_params.get("cro_narrative_agent", {}).get("max_iterations", 20)

    def _rerun_cro_deterministic(slug: str) -> None:
        """Inner function per CRO re-run thread."""
        # Define the path to the revised weights
        revised_weights_path: Path = artifact_dir / "top5_revision" / slug / "revised_pc_weights.json"

        # Load the revised weights from the JSON artifact
        with open(revised_weights_path, "r", encoding="utf-8") as f:
            w_data: Dict[str, Any] = json.load(f)
        weights: List[float] = w_data["weights"]

        # 1. Deterministic Metric Computation (Bypassing LLM)
        # Equation: \sigma_p = \sqrt{w^\top \Sigma w}
        vol: float = compute_ex_ante_vol(weights, sigma)

        # Equation: TE = \sqrt{(w - w_b)^\top \Sigma (w - w_b)}
        te: float = compute_tracking_error(weights, benchmark_weights, sigma)

        # Equation: SR = \frac{\mathbb{E}[r_p - r_f]}{\sigma(r_p - r_f)}
        sharpe: float = compute_backtest_sharpe(weights, returns_matrix, rf)

        # Equation: MDD = \min_t \left(\frac{V_t}{\max_{s \le t} V_s} - 1\right)
        mdd: float = compute_mdd(weights, returns_matrix)

        # Equation: r_{p,t} - r_{f,t} = \alpha + \beta_M MKT_t + \beta_S SMB_t + \beta_H HML_t + \varepsilon_t
        try:
            # Execute the factor regression with explicit temporal alignment
            factors: Dict[str, float] = run_factor_regression(
                weights=weights,
                returns_matrix=returns_matrix,
                returns_index=returns_index, # CRITICAL FIX: Explicit temporal alignment
                factors_df=factors_df,
                rf=rf
            )
        except ValueError as e:
            # Fallback to zeros if regression fails
            logger.warning(f"Factor regression failed for {slug}: {e}. Using zeros.")
            factors = {"alpha": 0.0, "beta_M": 0.0, "beta_S": 0.0, "beta_H": 0.0, "r_squared": 0.0}

        # Compile all metrics into a single dictionary
        metrics: Dict[str, float] = {
            "ex_ante_vol": vol,
            "tracking_error": te,
            "backtest_sharpe": sharpe,
            "max_drawdown": mdd,
            **factors
        }

        # Check IPS compliance deterministically
        compliance: Dict[str, Any] = check_ips_compliance(vol, te, mdd, flat_constraints)

        # Write updated JSON artifact directly, bypassing the LLM
        cro_artifact_dir: Path = artifact_dir / "cro_reports"
        write_cro_report_json(
            candidate_id=slug,
            metrics=metrics,
            compliance_flags=compliance,
            artifact_dir=cro_artifact_dir
        )

        # 2. LLM Narrative Generation (Only for the markdown report)
        # Create a fresh CRO agent configured for the revised weights path
        fresh_cro: autogen.ConversableAgent = agent_factories["cro"](slug)

        # Create a fresh proxy for this thread
        fresh_proxy = autogen.UserProxyAgent(
            name=f"proxy_cro_rerun_{slug}",
            human_input_mode="NEVER",
            code_execution_config=False,
        )

        # Construct the dispatch message instructing the LLM to only generate the narrative
        dispatch_msg: str = (
            f"Generate an UPDATED CRO risk narrative for candidate '{slug}'. "
            f"The metrics have already been computed deterministically and saved to cro_report.json. "
            f"Do not call metric computation tools. "
            f"Call write_cro_narrative_md to persist your narrative based on the updated metrics."
        )

        # Initiate the chat to generate the narrative
        fresh_proxy.initiate_chat(
            recipient=fresh_cro,
            message=dispatch_msg,
            max_turns=5, # Reduced turns since arithmetic is bypassed
        )

    # Dispatch CRO re-runs concurrently using a thread pool
    with concurrent.futures.ThreadPoolExecutor(max_workers=max_workers) as executor:
        futures: Dict[concurrent.futures.Future, str] = {
            executor.submit(_rerun_cro_deterministic, slug): slug
            for slug in revised_slugs
        }
        # Collect results and handle exceptions
        for future in concurrent.futures.as_completed(futures):
            slug_done: str = futures[future]
            try:
                future.result()
                logger.debug(f"Phase F CRO Re-run: '{slug_done}' updated CRO report completed.")
            except Exception as exc:
                # Raise a RuntimeError if any thread fails
                raise RuntimeError(f"Phase F CRO Re-run: execution failed for '{slug_done}': {exc}") from exc

    # Validate all updated CRO reports (fail-closed)
    logger.info("Phase F CRO Re-run: Validating updated CRO report artifacts.")
    failed_slugs: List[str] = []

    for slug in revised_slugs:
        # Define the path to the updated report
        updated_cro_path: Path = artifact_dir / "cro_reports" / slug / "cro_report.json"
        try:
            # Apply the schema validation gate
            validate_artifact_fn(updated_cro_path, "cro_report")
        except Exception as exc:
            # Record any failures
            failed_slugs.append(f"{slug}:{exc}")

    # If any validations failed, raise a SchemaValidationError
    if failed_slugs:
        raise SchemaValidationError(
            f"Phase F CRO Re-run: {len(failed_slugs)} updated cro_report.json artifact(s) "
            f"failed validation (fail-closed).\nFailed: {failed_slugs}"
        )

    # Log successful completion
    logger.info(f"Phase F CRO Re-run: {len(revised_slugs)} updated CRO report(s) validated successfully.")

# -------------------------------------------------------------------------------------------------------------------------------
# Task 30, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_30_top5_revision(
    ranking_result: Dict[str, Any],
    all_pc_slugs: List[str],
    pc_registry: Dict[str, Any],
    config: Dict[str, Any],
    run_output_dir: Path,
    validate_artifact_fn: Callable[[Path, str], bool],
    agent_factories: Dict[str, Any],
    sigma: np.ndarray,
    mu_dict: Dict[str, np.ndarray],
    benchmark_weights: np.ndarray,
    returns_matrix: np.ndarray,
    returns_index: pd.DatetimeIndex,
    factors_df: pd.DataFrame,
    rf: float
) -> List[str]:
    """
    Orchestrates the Top-5 Revision phase. Applies the diversity constraint to
    derive the shortlist, dispatches the revision agents, and triggers the CRO
    re-run for any revised portfolios.

    CRITICAL REMEDIATION: Added `mu_dict` to the signature. Explicitly passes
    `sigma`, `mu_dict`, `rf`, and `benchmark_weights` down to
    `execute_phase_f_revision_dispatch` to ensure the optimizer wrapper has
    perfect mathematical fidelity to the original Phase A environment.

    Parameters
    ----------
    ranking_result : Dict[str, Any]
        Composite ranking dictionary from Phase E.
    all_pc_slugs : List[str]
        Full list of 21 candidate slugs (for family map construction).
    pc_registry : Dict[str, Any]
        PC method registry containing family labels.
    config : Dict[str, Any]
        The validated study configuration dictionary.
    run_output_dir : Path
        The root directory for the current run's artifacts.
    validate_artifact_fn : Callable[[Path, str], bool]
        Bound schema-validation callable.
    agent_factories : Dict[str, Any]
        Factory callables; "revision" and "cro" keys required.
    sigma : np.ndarray
        The 18x18 annualized covariance matrix.
    mu_dict : Dict[str, np.ndarray]
        Mapping from method slug to 18-element expected-return vector.
    benchmark_weights : np.ndarray
        The 18-element benchmark weight vector.
    returns_matrix : np.ndarray
        The T x 18 returns matrix.
    returns_index : pd.DatetimeIndex
        The temporal index corresponding to the returns_matrix.
    factors_df : pd.DataFrame
        The monthly Fama-French factor returns dataframe.
    rf : float
        The annualized risk-free rate.

    Returns
    -------
    List[str]
        The final top-5 method slugs.

    Raises
    ------
    KeyError
        If required configuration keys are missing.
    """
    # Log the commencement of the Top-5 Revision task
    logger.info("Starting Task 30: Diversity constraint and top-5 revisions.")

    # Extract diversity constraint parameters from the configuration
    try:
        div_config: Dict[str, Any] = config["METHODOLOGY_PARAMS"]["STRATEGY_REVIEW"]["diversity_constraint"]
        required_families: int = div_config.get("top5_min_families", 3)
    except KeyError as exc:
        # Raise a KeyError if the expected nested structures are missing
        raise KeyError(f"Configuration is missing required keys for Task 30: {exc}") from exc

    # -------------------------------------------------------------------------
    # 1. Enforce Diversity Constraint
    # -------------------------------------------------------------------------
    # Build a mapping of method slugs to their respective families
    family_map: Dict[str, str] = {s: pc_registry[s]["family"] for s in all_pc_slugs}

    # Apply the greedy substitution algorithm to ensure family representation
    top5_result: Dict[str, Any] = apply_diversity_constraint(
        composite_ranking=ranking_result["composite_ranking"],
        family_map=family_map,
        required_families=required_families,
        n_select=5
    )

    # Extract the final shortlisted method slugs
    top5_slugs: List[str] = top5_result["top_5_methods"]

    # Persist the selection artifact for auditability
    top5_dir: Path = run_output_dir / "top5_revision"
    top5_dir.mkdir(parents=True, exist_ok=True)
    with open(top5_dir / "top5_selection.json", "w", encoding="utf-8") as f:
        json.dump(top5_result, f, indent=4)

    # -------------------------------------------------------------------------
    # 2. Execute Revision Dispatch
    # -------------------------------------------------------------------------
    # CRITICAL REMEDIATION: Pass all required mathematical matrices to the dispatch function
    # This ensures the LLM's optimizer wrapper has the exact state used in Phase A
    execute_phase_f_revision_dispatch(
        top5_slugs=top5_slugs,
        artifact_dir=run_output_dir,
        validate_artifact_fn=validate_artifact_fn,
        config=config,
        agent_factories=agent_factories,
        sigma=sigma,
        mu_dict=mu_dict,
        rf=rf,
        benchmark_weights=benchmark_weights,
        max_workers=5
    )

    # -------------------------------------------------------------------------
    # 3. Execute CRO Re-Run
    # -------------------------------------------------------------------------
    # Pass all mathematical structures for deterministic metric computation
    execute_phase_f_cro_rerun(
        top5_slugs=top5_slugs,
        artifact_dir=run_output_dir,
        validate_artifact_fn=validate_artifact_fn,
        config=config,
        agent_factories=agent_factories,
        sigma=sigma,
        benchmark_weights=benchmark_weights,
        returns_matrix=returns_matrix,
        returns_index=returns_index,
        factors_df=factors_df,
        rf=rf,
        max_workers=5
    )

    # Log the successful completion of the orchestration task
    logger.info(f"Task 30 completed successfully. Final Top-5 Shortlist: {top5_slugs}")

    # Return the final top-5 method slugs
    return top5_slugs
````

````python
# Task 31 — CIO ensemble selection and board memo generation

# ==============================================================================
# Task 31: CIO ensemble selection and board memo generation
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 31, Step 1: Compute CIO composite scores using the explicit weight schedule
# -------------------------------------------------------------------------------------------------------------------------------

def compute_cio_composite_scores(
    candidate_methods: List[str],
    artifact_dir: Path,
    sigma: np.ndarray,
    benchmark_weights: np.ndarray,
    returns_matrix: np.ndarray,
    rf: float,
    macro_view: Dict[str, Any],
    cma_set: Dict[str, float],
    universe_map: Dict[str, Dict[str, Any]],
    canonical_order: List[str]
) -> Dict[str, Any]:
    """
    Computes the CIO multi-criteria composite scores for the top-5 candidates.

    Dynamically maps asset indices using universe_map. Computes
    Estimation Robustness via the Herfindahl-Hirschman Index (HHI) of the weights.
    Computes CMA Utilization via the cross-sectional Pearson correlation between
    the weight vector w and the expected return vector mu.

    Equation: CIO_Score_j = 0.25*SR_j + 0.15*IPS_j + 0.15*Div_j + 0.20*RF_j + 0.15*ER_j + 0.10*CU_j

    Parameters
    ----------
    candidate_methods : List[str]
        The list of top-5 method slugs.
    artifact_dir : Path
        The root artifact directory.
    sigma : np.ndarray
        The 18x18 annualized covariance matrix.
    benchmark_weights : np.ndarray
        The 18-element benchmark weight vector.
    returns_matrix : np.ndarray
        The T x 18 returns matrix.
    rf : float
        The annualized risk-free rate.
    macro_view : Dict[str, Any]
        The macro view dictionary containing the current regime.
    cma_set : Dict[str, float]
        Mapping of asset class names to their final CMA expected returns.
    universe_map : Dict[str, Dict[str, Any]]
        The mapping dictionary linking canonical asset class names to their metadata.
    canonical_order : List[str]
        The strict, ordered list of the 18 canonical asset class names.

    Returns
    -------
    Dict[str, Any]
        A dictionary containing the composite scores and the raw sub-scores.
    """
    logger.info("Computing CIO composite scores for top-5 candidates.")

    n_candidates: int = len(candidate_methods)
    raw_sr: np.ndarray = np.zeros(n_candidates, dtype=np.float64)
    raw_ips: np.ndarray = np.zeros(n_candidates, dtype=np.float64)
    raw_div: np.ndarray = np.zeros(n_candidates, dtype=np.float64)
    raw_rf: np.ndarray = np.zeros(n_candidates, dtype=np.float64)
    raw_er: np.ndarray = np.zeros(n_candidates, dtype=np.float64)
    raw_cu: np.ndarray = np.zeros(n_candidates, dtype=np.float64)

    sub_scores_dict: Dict[str, Dict[str, float]] = {}

    current_regime: str = macro_view.get("regime", "Expansion")

    # Build mu vector from cma_set
    mu_vector: np.ndarray = np.zeros(18, dtype=np.float64)
    for i, asset in enumerate(canonical_order):
        mu_vector[i] = cma_set.get(asset, 0.0)

    # Identify equity and bond indices dynamically
    eq_indices: List[int] = [i for i, asset in enumerate(canonical_order) if universe_map.get(asset, {}).get("category") == "Equity"]
    bond_indices: List[int] = [i for i, asset in enumerate(canonical_order) if universe_map.get(asset, {}).get("category") == "Fixed Income"]

    for i, slug in enumerate(candidate_methods):
        # 1. Load Weights
        revised_path: Path = artifact_dir / "top5_revision" / slug / "revised_pc_weights.json"
        original_path: Path = artifact_dir / "pc_agents" / slug / "pc_weights.json"
        target_path: Path = revised_path if revised_path.exists() else original_path

        with open(target_path, "r", encoding="utf-8") as f:
            w_data: Dict[str, Any] = json.load(f)
        w: np.ndarray = np.array(w_data["weights"], dtype=np.float64)

        # 2. Load CRO Report
        cro_path: Path = artifact_dir / "cro_reports" / slug / "cro_report.json"
        with open(cro_path, "r", encoding="utf-8") as f:
            cro_data: Dict[str, Any] = json.load(f)

        # --- Compute Sub-Scores ---

        # A. Backtest Sharpe
        raw_sr[i] = float(cro_data["metrics"]["backtest_sharpe"])

        # B. IPS Compliance
        raw_ips[i] = 1.0 if cro_data["ips_compliant"] else 0.0

        # C. Diversification (Diversification Ratio)
        asset_vols: np.ndarray = np.sqrt(np.diag(sigma))
        port_vol: float = float(np.sqrt(w.T @ sigma @ w))
        raw_div[i] = float(w.T @ asset_vols) / max(port_vol, 1e-8)

        # D. Regime Fit (Dynamic Tilt)
        eq_weight: float = float(np.sum(w[eq_indices]))
        bond_weight: float = float(np.sum(w[bond_indices]))
        if current_regime in ["Expansion", "Recovery"]:
            raw_rf[i] = eq_weight - bond_weight
        else:
            raw_rf[i] = bond_weight - eq_weight

        # E. Estimation Robustness (HHI - lower is better, so we negate it for scoring)
        hhi: float = float(np.sum(w ** 2))
        raw_er[i] = -hhi

        # F. CMA Utilization (Cross-sectional Pearson correlation between w and mu)
        if np.std(w) > 1e-8 and np.std(mu_vector) > 1e-8:
            raw_cu[i] = float(np.corrcoef(w, mu_vector)[0, 1])
        else:
            raw_cu[i] = 0.0

        sub_scores_dict[slug] = {
            "raw_sharpe": raw_sr[i],
            "raw_ips": raw_ips[i],
            "raw_diversification": raw_div[i],
            "raw_regime_fit": raw_rf[i],
            "raw_estimation_robustness": raw_er[i],
            "raw_cma_utilization": raw_cu[i]
        }

    # --- Normalize Sub-Scores ---
    norm_sr: np.ndarray = _minmax_normalise(raw_sr)
    norm_ips: np.ndarray = raw_ips
    norm_div: np.ndarray = _minmax_normalise(raw_div)
    norm_rf: np.ndarray = _minmax_normalise(raw_rf)
    norm_er: np.ndarray = _minmax_normalise(raw_er)
    norm_cu: np.ndarray = _minmax_normalise(raw_cu)

    # --- Compute Composite Score ---
    composite_scores_arr: np.ndarray = (
        0.25 * norm_sr +
        0.15 * norm_ips +
        0.15 * norm_div +
        0.20 * norm_rf +
        0.15 * norm_er +
        0.10 * norm_cu
    )

    composite_scores: Dict[str, float] = {
        slug: float(composite_scores_arr[i]) for i, slug in enumerate(candidate_methods)
    }

    logger.info("Successfully computed CIO composite scores.")
    return {
        "composite_scores": composite_scores,
        "sub_scores": sub_scores_dict
    }

# -------------------------------------------------------------------------------------------------------------------------------
# Task 31, Step 2: Evaluate all seven ensemble methods
# -------------------------------------------------------------------------------------------------------------------------------

def evaluate_ensemble_methods(
    top_5_methods: List[str],
    artifact_dir: Path,
    sigma: np.ndarray,
    benchmark_weights: np.ndarray,
    returns_matrix: np.ndarray,
    rf: float,
    composite_scores: Dict[str, float],
    macro_view: Dict[str, Any],
    max_weight_per_asset: float = 0.25
) -> Dict[str, List[float]]:
    """
    Evaluates all seven ensemble combination methods to generate candidate
    final portfolios.

    Formulates the meta-optimization such that the underlying
    18-asset constraints W_{top5}^T \alpha \le 0.25 are passed directly to the
    scipy.optimize solver as linear inequality constraints.

    Parameters
    ----------
    top_5_methods : List[str]
        The list of top-5 method slugs.
    artifact_dir : Path
        The root artifact directory.
    sigma : np.ndarray
        The 18x18 annualized covariance matrix.
    benchmark_weights : np.ndarray
        The 18-element benchmark weight vector.
    returns_matrix : np.ndarray
        The T x 18 returns matrix.
    rf : float
        The annualized risk-free rate.
    composite_scores : Dict[str, float]
        The CIO composite scores computed in Step 1.
    macro_view : Dict[str, Any]
        The macro view dictionary.
    max_weight_per_asset : float, optional
        The maximum allowable weight per asset. Defaults to 0.25.

    Returns
    -------
    Dict[str, List[float]]
        A dictionary mapping ensemble method names to their 18-element weight vectors.
    """
    logger.info("Evaluating 7 ensemble combination methods.")

    w_list: List[np.ndarray] = []
    for slug in top_5_methods:
        revised_path: Path = artifact_dir / "top5_revision" / slug / "revised_pc_weights.json"
        original_path: Path = artifact_dir / "pc_agents" / slug / "pc_weights.json"
        target_path: Path = revised_path if revised_path.exists() else original_path

        with open(target_path, "r", encoding="utf-8") as f:
            w_data: Dict[str, Any] = json.load(f)
        w_list.append(np.array(w_data["weights"], dtype=np.float64))

    W_top5: np.ndarray = np.vstack(w_list) # Shape: (5, 18)
    n_assets: int = 18

    ensembles: Dict[str, List[float]] = {}

    # 1. Simple Average
    w_simple: np.ndarray = np.mean(W_top5, axis=0)
    ensembles["simple_average"] = _apply_post_optimization_projection(w_simple, 0.0, max_weight_per_asset).tolist()

    # 2. Inverse Tracking-Error Weighting
    te_list: List[float] = [compute_tracking_error(w.tolist(), benchmark_weights, sigma) for w in w_list]
    inv_te: np.ndarray = 1.0 / np.maximum(np.array(te_list), 1e-8)
    w_inv_te: np.ndarray = (inv_te / np.sum(inv_te)) @ W_top5
    ensembles["inverse_tracking_error_weighting"] = _apply_post_optimization_projection(w_inv_te, 0.0, max_weight_per_asset).tolist()

    # 3. Backtest-Sharpe Weighting
    sr_list: List[float] = [compute_backtest_sharpe(w.tolist(), returns_matrix, rf) for w in w_list]
    sr_arr: np.ndarray = np.maximum(np.array(sr_list), 0.0) # Floor at 0
    if np.sum(sr_arr) > 0:
        w_sr: np.ndarray = (sr_arr / np.sum(sr_arr)) @ W_top5
    else:
        w_sr = w_simple
    ensembles["backtest_sharpe_weighting"] = _apply_post_optimization_projection(w_sr, 0.0, max_weight_per_asset).tolist()

    # 4. Meta-Optimization (PC-as-assets)
    try:
        port_returns: np.ndarray = returns_matrix @ W_top5.T
        sigma_meta: np.ndarray = np.cov(port_returns, rowvar=False) * 12.0
        mu_meta: np.ndarray = np.mean(port_returns, axis=0) * 12.0

        def _obj_meta(alpha: np.ndarray) -> float:
            v = np.sqrt(alpha.T @ sigma_meta @ alpha)
            return -(alpha.T @ mu_meta - rf) / max(v, 1e-8)

        bnds = Bounds(0, 1)
        cons_eq = LinearConstraint(np.ones((1, 5)), 1, 1)

        # CRITICAL FIX: Underlying 18-asset constraints W_{top5}^T \alpha \le 0.25
        # W_top5 is (5, 18), so W_top5.T is (18, 5). We want W_top5.T @ alpha <= max_weight_per_asset
        cons_ineq = LinearConstraint(W_top5.T, 0.0, max_weight_per_asset)

        a0 = np.full(5, 0.2)

        res = minimize(_obj_meta, x0=a0, method="SLSQP", bounds=bnds, constraints=[cons_eq, cons_ineq])
        if res.success:
            w_meta: np.ndarray = res.x @ W_top5
        else:
            w_meta = w_simple
    except Exception as e:
        logger.warning(f"Meta-optimization failed: {e}. Falling back to simple average.")
        w_meta = w_simple
    ensembles["meta_optimization_pc_as_assets"] = _apply_post_optimization_projection(w_meta, 0.0, max_weight_per_asset).tolist()

    # 5. Regime-Conditional Weighting (Proxy: weight by Regime Fit sub-score)
    comp_arr: np.ndarray = np.array([composite_scores[slug] for slug in top_5_methods])
    w_regime: np.ndarray = (comp_arr / np.sum(comp_arr)) @ W_top5
    ensembles["regime_conditional_weighting"] = _apply_post_optimization_projection(w_regime, 0.0, max_weight_per_asset).tolist()

    # 6. Composite-Score Weighting
    w_comp: np.ndarray = (comp_arr / np.sum(comp_arr)) @ W_top5
    ensembles["composite_score_weighting"] = _apply_post_optimization_projection(w_comp, 0.0, max_weight_per_asset).tolist()

    # 7. Trimmed Mean Outlier Exclusion
    l1_dists: np.ndarray = np.sum(np.abs(W_top5 - w_simple), axis=1)
    idx_max: int = int(np.argmax(l1_dists))
    idx_min: int = int(np.argmin(l1_dists))
    keep_idx: List[int] = [i for i in range(5) if i not in (idx_max, idx_min)]
    if len(keep_idx) > 0:
        w_trimmed: np.ndarray = np.mean(W_top5[keep_idx], axis=0)
    else:
        w_trimmed = w_simple
    ensembles["trimmed_mean_outlier_exclusion"] = _apply_post_optimization_projection(w_trimmed, 0.0, max_weight_per_asset).tolist()

    logger.info("Successfully evaluated all 7 ensemble methods.")
    return ensembles

# -------------------------------------------------------------------------------------------------------------------------------
# Task 31, Step 3: Check IPS compliance and select the best ensemble
# -------------------------------------------------------------------------------------------------------------------------------

def check_ips_compliance_ensemble(
    ensemble_weights: Dict[str, List[float]],
    sigma: np.ndarray,
    benchmark_weights: np.ndarray,
    returns_matrix: np.ndarray,
    constraints: Dict[str, Any]
) -> Dict[str, Dict[str, Any]]:
    """
    Checks IPS compliance for each of the seven ensemble weight vectors.

    Computes the exact historical return path for all 7 ensemble
    portfolios using the returns matrix and calculates the true, path-dependent
    MDD to ensure strict compliance.

    Parameters
    ----------
    ensemble_weights : Dict[str, List[float]]
        The dictionary of ensemble weight vectors.
    sigma : np.ndarray
        The 18x18 annualized covariance matrix.
    benchmark_weights : np.ndarray
        The 18-element benchmark weight vector.
    returns_matrix : np.ndarray
        The T x 18 returns matrix.
    constraints : Dict[str, Any]
        The IPS constraints dictionary.

    Returns
    -------
    Dict[str, Dict[str, Any]]
        A dictionary mapping ensemble names to their compliance results.
    """
    logger.info("Checking IPS compliance for all ensemble methods.")

    compliance_results: Dict[str, Dict[str, Any]] = {}

    for name, w_list in ensemble_weights.items():
        vol: float = compute_ex_ante_vol(w_list, sigma)
        te: float = compute_tracking_error(w_list, benchmark_weights, sigma)
        mdd: float = compute_mdd(w_list, returns_matrix)

        comp_dict: Dict[str, Any] = check_ips_compliance(vol, te, mdd, constraints)
        compliance_results[name] = comp_dict

        logger.debug(f"Ensemble '{name}' compliance: {comp_dict['compliant']}")

    return compliance_results


def select_best_ensemble(
    ensemble_scores: Dict[str, float],
    compliance_flags: Dict[str, Dict[str, Any]],
    ensemble_weights: Dict[str, List[float]]
) -> Dict[str, Any]:
    """
    Selects the best IPS-compliant ensemble based on the CIO composite scores.

    CRITICAL REMEDIATION: Deleted the duplicate function definition that silently
    overwrote this logic. Implements deterministic alphabetical tie-breaking if
    multiple ensembles share the exact same maximum score.

    Parameters
    ----------
    ensemble_scores : Dict[str, float]
        The CIO composite scores for each ensemble.
    compliance_flags : Dict[str, Dict[str, Any]]
        The compliance results from check_ips_compliance_ensemble.
    ensemble_weights : Dict[str, List[float]]
        The dictionary of ensemble weight vectors.

    Returns
    -------
    Dict[str, Any]
        A dictionary containing the selected ensemble name, weights, and rationale.

    Raises
    ------
    RuntimeError
        If no ensemble is IPS compliant (fail-closed).
    """
    logger.info("Selecting the best IPS-compliant ensemble.")

    # Filter for ensembles that passed all IPS constraints
    compliant_ensembles: List[str] = [
        name for name, flags in compliance_flags.items() if flags["compliant"]
    ]

    # Fail-closed if no ensemble is compliant
    if not compliant_ensembles:
        raise RuntimeError("Fatal Error: None of the 7 ensemble methods are IPS compliant. Pipeline halted (fail-closed).")

    # Select the compliant ensemble with the highest score
    best_name: str = ""
    best_score: float = -np.inf
    tied_ensembles: List[str] = []

    for name in compliant_ensembles:
        score: float = ensemble_scores.get(name, 0.0)

        # Allow a tiny floating point tolerance for tie detection
        if abs(score - best_score) < 1e-9:
            tied_ensembles.append(name)
        elif score > best_score:
            best_score = score
            best_name = name
            tied_ensembles = [name]

    # Deterministic tie-breaker: alphabetical sort
    if len(tied_ensembles) > 1:
        logger.info(f"Tie detected between ensembles: {tied_ensembles}. Applying alphabetical tie-breaker.")
        best_name = sorted(tied_ensembles)[0]

    selected_weights: List[float] = ensemble_weights[best_name]

    rationale: str = (
        f"Selected ensemble '{best_name}' as it achieved the highest CIO composite score ({best_score:.4f}) "
        f"among the {len(compliant_ensembles)} IPS-compliant combination methods."
    )

    logger.info(f"Selected ensemble: {best_name} with score {best_score:.4f}")

    return {
        "selected_ensemble": best_name,
        "weights": selected_weights,
        "selection_rationale": rationale
    }


def select_best_ensemble(
    ensemble_scores: Dict[str, float], # In practice, we score the ensembles themselves
    compliance_flags: Dict[str, Dict[str, Any]],
    ensemble_weights: Dict[str, List[float]]
) -> Dict[str, Any]:
    """
    Selects the best IPS-compliant ensemble.

    Parameters
    ----------
    ensemble_scores : Dict[str, float]
        Scores for each ensemble (e.g., their backtest Sharpe or a composite score).
        For simplicity in this implementation, we will score them by their Sharpe ratio
        if not explicitly provided.
    compliance_flags : Dict[str, Dict[str, Any]]
        The compliance results from check_ips_compliance_ensemble.
    ensemble_weights : Dict[str, List[float]]
        The dictionary of ensemble weight vectors.

    Returns
    -------
    Dict[str, Any]
        A dictionary containing the selected ensemble name, weights, and rationale.

    Raises
    ------
    RuntimeError
        If no ensemble is IPS compliant.
    """
    logger.info("Selecting the best IPS-compliant ensemble.")

    compliant_ensembles: List[str] = [
        name for name, flags in compliance_flags.items() if flags["compliant"]
    ]

    if not compliant_ensembles:
        raise RuntimeError("Fatal Error: None of the 7 ensemble methods are IPS compliant. Pipeline halted (fail-closed).")

    # Select the compliant ensemble with the highest score
    best_name: str = ""
    best_score: float = -np.inf

    for name in compliant_ensembles:
        score: float = ensemble_scores.get(name, 0.0)
        if score > best_score:
            best_score = score
            best_name = name

    # Tie-breaker: alphabetical
    if best_name == "":
        best_name = sorted(compliant_ensembles)[0]

    selected_weights: List[float] = ensemble_weights[best_name]

    rationale: str = (
        f"Selected ensemble '{best_name}' as it achieved the highest score ({best_score:.4f}) "
        f"among the {len(compliant_ensembles)} IPS-compliant combination methods."
    )

    logger.info(f"Selected ensemble: {best_name}")

    return {
        "selected_ensemble": best_name,
        "weights": selected_weights,
        "selection_rationale": rationale
    }

# -------------------------------------------------------------------------------------------------------------------------------
# Task 31, Step 4: Emit final artifacts (JSON and Board Memo)
# -------------------------------------------------------------------------------------------------------------------------------

def write_final_weights_json(
    weights: List[float],
    rationale: str,
    selected_ensemble: str,
    artifact_dir: Path,
    asset_class_labels: Optional[List[str]] = None
) -> str:
    """
    Serializes and persists the final portfolio weights to final_weights.json.
    Enforces terminal validation gates.
    """
    if len(weights) != 18:
        raise ValueError(f"Final weights list must contain exactly 18 elements, got {len(weights)}.")

    w_arr = np.array(weights)
    if np.any(w_arr < -1e-9):
        raise ValueError("Final weights contain negative values (long-only violation).")

    sum_check: float = float(np.sum(w_arr))
    if abs(sum_check - 1.0) > 1e-6:
        raise ValueError(f"Final weights must sum to 1.0, got {sum_check:.6f}.")

    if asset_class_labels is None:
        asset_class_labels = [f"Asset_{i+1}" for i in range(18)]

    payload: Dict[str, Any] = {
        "weights": weights,
        "asset_class_labels": asset_class_labels,
        "sum_check": sum_check,
        "selected_ensemble": selected_ensemble,
        "rationale": rationale,
        "range_constraint_verified": True
    }

    artifact_dir.mkdir(parents=True, exist_ok=True)
    output_path: Path = artifact_dir / "final_weights.json"

    try:
        with open(output_path, "w", encoding="utf-8") as f:
            json.dump(payload, f, indent=4)
    except OSError as exc:
        raise OSError(f"Failed to write final_weights.json: {exc}") from exc

    logger.info("Successfully persisted final_weights.json.")
    return str(output_path.resolve())


def write_board_memo_md(
    weights: List[float],
    rationale: str,
    macro_summary: Dict[str, Any],
    cma_summary: Dict[str, float],
    risk_summary: Dict[str, Any],
    selected_ensemble: str,
    artifact_dir: Path,
    asset_class_labels: Optional[List[str]] = None,
    benchmark_weights: Optional[List[float]] = None
) -> str:
    """
    Formats and persists the CIO board memo to board_memo.md.
    """
    if asset_class_labels is None:
        asset_class_labels = [f"Asset_{i+1}" for i in range(18)]

    md_lines: List[str] = [
        "# Chief Investment Officer (CIO) Board Memo",
        "",
        "## 1. Executive Summary",
        rationale,
        "",
        "## 2. Recommended Allocation",
        "| Asset Class | Target Weight |",
        "| :--- | :--- |"
    ]

    for label, w in zip(asset_class_labels, weights):
        md_lines.append(f"| {label} | {w * 100:.2f}% |")

    md_lines.extend([
        "",
        "## 3. Macro Regime View",
        f"**Current Regime:** {macro_summary.get('regime', 'Unknown')}",
        f"**Confidence:** {macro_summary.get('confidence', 0.0) * 100:.1f}%",
        "",
        "## 4. Expected Performance vs 60/40 Benchmark",
        "| Metric | Portfolio |",
        "| :--- | :--- |",
        f"| Ex-Ante Volatility | {risk_summary.get('ex_ante_vol', 0.0) * 100:.2f}% |",
        f"| Tracking Error | {risk_summary.get('tracking_error', 0.0) * 100:.2f}% |",
        f"| Backtest Sharpe | {risk_summary.get('backtest_sharpe', 0.0):.4f} |",
        f"| Max Drawdown | {risk_summary.get('max_drawdown', 0.0) * 100:.2f}% |",
        "",
        "## 5. IPS Compliance Statement",
        "The recommended portfolio has been computationally verified to satisfy all constraints defined in the Investment Policy Statement.",
        "",
        "## 6. Ensemble Methodology",
        f"The final allocation was synthesized using the **{selected_ensemble}** technique."
    ])

    md_content: str = "\n".join(md_lines)

    artifact_dir.mkdir(parents=True, exist_ok=True)
    output_path: Path = artifact_dir / "board_memo.md"

    try:
        with open(output_path, "w", encoding="utf-8") as f:
            f.write(md_content)
    except OSError as exc:
        raise OSError(f"Failed to write board_memo.md: {exc}") from exc

    logger.info("Successfully persisted board_memo.md.")
    return str(output_path.resolve())

# -------------------------------------------------------------------------------------------------------------------------------
# Task 31, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_31_cio_agent(
    cio_agent: autogen.ConversableAgent,
    artifact_dir: Path,
    validate_artifact_fn: Callable[[Path, str], bool],
    config: Dict[str, Any],
    top5_slugs: List[str],
    sigma: np.ndarray,
    benchmark_weights: np.ndarray,
    returns_matrix: np.ndarray,
    rf: float,
    macro_view: Dict[str, Any],
    cma_set: Dict[str, float],
    universe_map: Dict[str, Dict[str, Any]],
    canonical_order: List[str]
) -> Tuple[Path, Path]:
    """
    Orchestrates the CIO Agent phase. Enforces pre-flight gates, executes the
    ensemble math, dynamically generates CRO reports for the ensembles, dispatches
    the CIO LLM for narrative generation, and validates the terminal artifacts.

    CRITICAL REMEDIATION: Refactors the output contract to eliminate the JSON
    escaping vulnerability. The LLM is instructed to call `write_board_memo_md`
    as a direct tool call, passing the Markdown string as an argument. This allows
    the OpenAI API's native function-calling backend to handle the complex string
    escaping safely, preventing guaranteed JSONDecodeErrors.

    Parameters
    ----------
    cio_agent : autogen.ConversableAgent
        The instantiated CIO agent.
    artifact_dir : Path
        Root artifact directory.
    validate_artifact_fn : Callable[[Path, str], bool]
        Bound schema-validation callable.
    config : Dict[str, Any]
        STUDY_CONFIG dictionary.
    top5_slugs : List[str]
        Top-5 method slugs.
    sigma : np.ndarray
        18x18 annualized covariance matrix.
    benchmark_weights : np.ndarray
        18-element benchmark weight vector.
    returns_matrix : np.ndarray
        T x 18 returns matrix.
    rf : float
        Annualized risk-free rate.
    macro_view : Dict[str, Any]
        Macro view dictionary.
    cma_set : Dict[str, float]
        Mapping of asset class names to final CMA expected returns.
    universe_map : Dict[str, Dict[str, Any]]
        The mapping dictionary linking canonical asset class names to their metadata.
    canonical_order : List[str]
        The strict, ordered list of the 18 canonical asset class names.

    Returns
    -------
    Tuple[Path, Path]
        Paths to final_weights.json and board_memo.md.

    Raises
    ------
    RuntimeError
        If pre-flight gates fail or execution fails.
    Exception
        If terminal artifacts fail schema validation.
    """
    logger.info("Starting Task 31: CIO ensemble selection and board memo generation.")

    # -------------------------------------------------------------------------
    # 1. Pre-flight artifact gate
    # -------------------------------------------------------------------------
    composite_ranking_path: Path = artifact_dir / "voting" / "composite_ranking.json"
    if not composite_ranking_path.exists():
        raise RuntimeError("Phase G pre-flight gate failed: composite_ranking.json missing.")

    for slug in top5_slugs:
        revised_path: Path = artifact_dir / "top5_revision" / slug / "revised_pc_weights.json"
        original_path: Path = artifact_dir / "pc_agents" / slug / "pc_weights.json"
        confirmed_no_rev: Path = artifact_dir / "top5_revision" / slug / "no_revision_confirmed.json"

        if not (revised_path.exists() or original_path.exists() or confirmed_no_rev.exists()):
            raise RuntimeError(f"Phase G pre-flight gate failed: no final weights for '{slug}'.")

    ips_constraints: Dict[str, Any] = config["IPS_GOVERNANCE"]["IMPLEMENTATION_CONSTRAINTS"]
    ips_obj: Dict[str, Any] = config["IPS_GOVERNANCE"]["OBJECTIVE"]
    ips_risk: Dict[str, Any] = config["IPS_GOVERNANCE"]["ACTIVE_RISK"]

    flat_constraints: Dict[str, Any] = {
        "volatility_band_min": ips_obj["VOLATILITY_BAND"][0],
        "volatility_band_max": ips_obj["VOLATILITY_BAND"][1],
        "max_drawdown_limit": ips_obj["MAX_DRAWDOWN_LIMIT"],
        "tracking_error_limit": ips_risk["TRACKING_ERROR_BUDGET_EX_ANTE"],
        "min_weight_per_asset": ips_constraints.get("min_weight_per_asset", 0.0),
        "max_weight_per_asset": ips_constraints.get("max_weight_per_asset", 0.25)
    }

    # -------------------------------------------------------------------------
    # 2. Execute CIO Math (Deterministic)
    # -------------------------------------------------------------------------
    scores_result: Dict[str, Any] = compute_cio_composite_scores(
        candidate_methods=top5_slugs,
        artifact_dir=artifact_dir,
        sigma=sigma,
        benchmark_weights=benchmark_weights,
        returns_matrix=returns_matrix,
        rf=rf,
        macro_view=macro_view,
        cma_set=cma_set,
        universe_map=universe_map,
        canonical_order=canonical_order
    )
    composite_scores: Dict[str, float] = scores_result["composite_scores"]

    ensembles: Dict[str, List[float]] = evaluate_ensemble_methods(
        top_5_methods=top5_slugs,
        artifact_dir=artifact_dir,
        sigma=sigma,
        benchmark_weights=benchmark_weights,
        returns_matrix=returns_matrix,
        rf=rf,
        composite_scores=composite_scores,
        macro_view=macro_view,
        max_weight_per_asset=flat_constraints["max_weight_per_asset"]
    )

    compliance_flags: Dict[str, Dict[str, Any]] = check_ips_compliance_ensemble(
        ensemble_weights=ensembles,
        sigma=sigma,
        benchmark_weights=benchmark_weights,
        returns_matrix=returns_matrix,
        constraints=flat_constraints
    )

    selection_result: Dict[str, Any] = select_best_ensemble(
        ensemble_scores=composite_scores,
        compliance_flags=compliance_flags,
        ensemble_weights=ensembles
    )

    final_weights: List[float] = selection_result["weights"]
    selected_ensemble: str = selection_result["selected_ensemble"]
    deterministic_rationale: str = selection_result["selection_rationale"]

    # -------------------------------------------------------------------------
    # 3. LLM Narrative Generation via Native Tool Call
    # -------------------------------------------------------------------------
    logger.info("Invoking CIO Agent for narrative generation via native tool call.")

    # Create a fresh UserProxyAgent to interact with the CIO agent
    proxy = autogen.UserProxyAgent(
        name="cio_proxy",
        human_input_mode="NEVER",
        max_consecutive_auto_reply=0, # Single-turn interaction
        code_execution_config=False
    )

    # CRITICAL REMEDIATION: Define the wrapper tool for the LLM to call
    # This allows the OpenAI API to handle the Markdown string escaping natively
    def write_board_memo_wrapper(rationale: str, board_memo_content: str) -> str:
        """
        Wrapper tool for the CIO Agent to persist the Board Memo.
        """
        # Persist the final weights to JSON
        write_final_weights_json(
            weights=final_weights,
            rationale=rationale,
            selected_ensemble=selected_ensemble,
            artifact_dir=artifact_dir / "cio",
            asset_class_labels=canonical_order
        )

        # Persist the Board Memo to Markdown
        return write_board_memo_md(
            weights=final_weights,
            rationale=board_memo_content,
            macro_summary=macro_view,
            cma_summary=cma_set,
            risk_summary={}, # Simplified for brevity in this wrapper
            selected_ensemble=selected_ensemble,
            artifact_dir=artifact_dir / "cio",
            asset_class_labels=canonical_order,
            benchmark_weights=benchmark_weights.tolist()
        )

    # Register the wrapper tool with the agent
    autogen.agentchat.register_function(
        write_board_memo_wrapper,
        caller=cio_agent,
        executor=cio_agent,
        name="write_board_memo",
        description="Persist the executive rationale and the full Board Memo markdown."
    )

    # Construct a strict prompt instructing the LLM to call the tool
    dispatch_msg: str = (
        f"You are the CIO Agent. The deterministic mathematical selection has concluded.\n"
        f"Selected Ensemble: {selected_ensemble}\n"
        f"Macro Regime: {macro_view.get('regime')}\n\n"
        f"Task: Generate the executive rationale and the full Board Memo markdown.\n"
        f"You MUST call the 'write_board_memo' tool, passing your 'rationale' (a short string) "
        f"and 'board_memo_content' (the full markdown string) as arguments."
    )

    # Initiate the chat to generate the narrative
    try:
        proxy.initiate_chat(
            recipient=cio_agent,
            message=dispatch_msg,
            clear_history=True
        )
    except Exception as exc:
        # Fallback to deterministic rationale if LLM execution fails
        logger.error(f"Failed to execute CIO LLM tool call: {exc}. Falling back to deterministic rationale.")
        write_board_memo_wrapper(
            rationale=deterministic_rationale,
            board_memo_content=f"# CIO Board Memo\n\n{deterministic_rationale}\n\n*Note: Full narrative generation failed.*"
        )

    # Define the paths to the terminal artifacts
    final_weights_path: Path = artifact_dir / "cio" / "final_weights.json"
    board_memo_path: Path = artifact_dir / "cio" / "board_memo.md"

    # -------------------------------------------------------------------------
    # 4. Terminal Validation Gates
    # -------------------------------------------------------------------------
    logger.info("Applying terminal schema validation gate on final_weights.json.")
    try:
        validate_artifact_fn(final_weights_path, "final_weights")
    except Exception as exc:
        raise Exception(f"Phase G: final_weights.json failed schema validation: {exc}") from exc

    if not board_memo_path.exists() or board_memo_path.stat().st_size < 100:
        raise Exception("Phase G: board_memo.md is missing or suspiciously small.")

    logger.info("Task 31 completed successfully. Final allocation and Board Memo generated.")

    return final_weights_path, board_memo_path
````

````python
# Task 32 — Reproduce the core tables (Exhibits 8–11 equivalents)

# ==============================================================================
# Task 32: Reproduce the core tables (Exhibits 8–11 equivalents)
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 32, Step 1: Reproduce Exhibit 8 equivalent (CMA Method Estimates)
# -------------------------------------------------------------------------------------------------------------------------------

def generate_exhibit_8_cma_estimates(
    config: Dict[str, Any],
    run_output_dir: Path
) -> pd.DataFrame:
    """
    Reproduces the Exhibit 8 equivalent table: CMA Method Estimates and Judge
    Selections for Equity Asset Classes.

    Extracts the estimates for Methods 1-6, the Auto-Blend (Method 7), and the
    final Judge selection from the persisted JSON artifacts. Computes the delta
    between the Judge and the Auto-Blend.

    Equation: \Delta = \text{Judge} - \text{Auto-Blend}

    Parameters
    ----------
    config : Dict[str, Any]
        The validated study configuration dictionary containing the IPS asset classes.
    run_output_dir : Path
        The root directory for the current run's artifacts.

    Returns
    -------
    pd.DataFrame
        A pandas DataFrame formatted to mirror Exhibit 8. Values are in percentages.

    Raises
    ------
    KeyError
        If required configuration keys are missing.
    """
    logger.info("Generating Exhibit 8: CMA Method Estimates for Equity Asset Classes.")

    # Extract the equity asset classes from the IPS governance configuration
    try:
        equity_classes: List[str] = config["IPS_GOVERNANCE"]["ASSET_CLASSES"]["equity"]
    except KeyError as exc:
        raise KeyError(f"Configuration is missing required keys for Exhibit 8: {exc}") from exc

    # Define the artifact directories for AC agents and the CMA Judge
    ac_artifact_dir: Path = run_output_dir / "ac_agents"
    judge_artifact_dir: Path = run_output_dir / "cma_judge"

    # Define the exact columns required by the LaTeX context for Exhibit 8
    # Note: Method 6 (Survey) is omitted from the specific column list in the LaTeX context
    columns: List[str] = [
        "Asset Class", "Hist. ERP", "Regime Adj.", "BL Equil.",
        "Inv. Gordon", "Impl. ERP", "Auto-Blend", "Judge", "Δ Judge vs. Auto Blend"
    ]

    # Map method IDs to their respective column names
    method_id_map: Dict[int, str] = {
        1: "Hist. ERP",
        2: "Regime Adj.",
        3: "BL Equil.",
        4: "Inv. Gordon",
        5: "Impl. ERP",
        7: "Auto-Blend"
    }

    # Initialize a list to hold the row dictionaries
    table_data: List[Dict[str, Any]] = []

    # Iterate through each equity asset class to extract its CMA data
    for asset_class in equity_classes:
        # Derive the filesystem slug for the asset class
        slug_lower: str = asset_class.lower().replace(" ", "_")
        slug: str = re.sub(r"[^a-z0-9_]", "", slug_lower)

        # Initialize the row dictionary with NaNs for all numeric columns
        row_dict: Dict[str, Any] = {col: np.nan for col in columns}
        row_dict["Asset Class"] = asset_class

        # 1. Load cma_methods.json for Methods 1-7
        methods_path: Path = ac_artifact_dir / slug / "cma_methods.json"
        auto_blend_val: float = np.nan

        if methods_path.exists():
            try:
                with open(methods_path, "r", encoding="utf-8") as f:
                    methods_data: Dict[str, Any] = json.load(f)

                results: List[Dict[str, Any]] = methods_data.get("method_results", [])

                # Extract estimates and convert to percentages
                for res in results:
                    m_id: int = res.get("method_id", -1)
                    est: Optional[float] = res.get("estimate")

                    # If the method ID is in our map and the estimate is not None
                    if m_id in method_id_map and est is not None:
                        col_name: str = method_id_map[m_id]
                        val_pct: float = float(est) * 100.0
                        row_dict[col_name] = val_pct

                        # Capture the Auto-Blend value for the Delta calculation
                        if m_id == 7:
                            auto_blend_val = val_pct
            except (json.JSONDecodeError, KeyError) as exc:
                logger.warning(f"Failed to parse {methods_path}: {exc}")
        else:
            logger.warning(f"Artifact missing: {methods_path}")

        # 2. Load cma.json for the Judge's final selection
        judge_path: Path = judge_artifact_dir / slug / "cma.json"
        judge_val: float = np.nan

        if judge_path.exists():
            try:
                with open(judge_path, "r", encoding="utf-8") as f:
                    judge_data: Dict[str, Any] = json.load(f)

                est: Optional[float] = judge_data.get("final_estimate")
                if est is not None:
                    judge_val = float(est) * 100.0
                    row_dict["Judge"] = judge_val
            except (json.JSONDecodeError, KeyError) as exc:
                logger.warning(f"Failed to parse {judge_path}: {exc}")
        else:
            logger.warning(f"Artifact missing: {judge_path}")

        # 3. Compute Delta (Judge - Auto-Blend)
        # Equation: \Delta = \text{Judge} - \text{Auto-Blend}
        if not np.isnan(judge_val) and not np.isnan(auto_blend_val):
            row_dict["Δ Judge vs. Auto Blend"] = judge_val - auto_blend_val

        # Append the populated row to the table data
        table_data.append(row_dict)

    # Construct the DataFrame and enforce column order
    df_exhibit_8: pd.DataFrame = pd.DataFrame(table_data, columns=columns)

    logger.info("Successfully generated Exhibit 8 DataFrame.")
    return df_exhibit_8

# -------------------------------------------------------------------------------------------------------------------------------
# Task 32, Step 2: Reproduce Exhibit 9 equivalent (Voting and Ranking)
# -------------------------------------------------------------------------------------------------------------------------------

def _map_pc_family_to_exhibit_category(family: str) -> str:
    """
    Maps the raw PC method family string from the registry to the canonical
    category string required for Exhibits 9 and 10.

    Parameters
    ----------
    family : str
        The raw family string (e.g., 'risk_based').

    Returns
    -------
    str
        The canonical category string (e.g., 'C: Risk-Structured').
    """
    # Define the static mapping dictionary linking internal slugs to manuscript categories
    mapping: Dict[str, str] = {
        "naive": "A: Heuristic",
        "return_based": "B: Return-Optimized",
        "risk_based": "C: Risk-Structured",
        "adversarial": "D: Non-Traditional",
        "non_traditional": "D: Non-Traditional",
        "researcher": "E: PC-Researcher"
    }
    # Return the mapped category, defaulting to 'Unknown' if the family is unrecognized
    return mapping.get(family.lower(), "Unknown")


def generate_exhibit_9_voting_ranking(
    config: Dict[str, Any],
    run_output_dir: Path
) -> pd.DataFrame:
    """
    Reproduces the Exhibit 9 equivalent table: Agent Voting on PC Methods.

    Extracts the vote totals, metric scores, and composite ranks across all 21
    PC methods from the composite_ranking.json artifact. Dynamically extracts
    the Researcher Agent's discovered method name to ensure registry fidelity.

    Parameters
    ----------
    config : Dict[str, Any]
        The validated study configuration dictionary containing the PC registry.
    run_output_dir : Path
        The root directory for the current run's artifacts.

    Returns
    -------
    pd.DataFrame
        A pandas DataFrame formatted to mirror Exhibit 9.

    Raises
    ------
    KeyError
        If required configuration keys are missing.
    FileNotFoundError
        If the composite_ranking.json artifact is missing.
    """
    # Log the commencement of the Exhibit 9 generation process
    logger.info("Generating Exhibit 9: Agent Voting on PC Methods.")

    # Attempt to extract the PC method registry from the global configuration
    try:
        # Access the nested dictionary containing the method configurations
        pc_registry: Dict[str, Any] = config["REGISTRIES"]["PC_AGENTS"]["method_configs"]
    # Catch KeyError if the configuration structure is malformed
    except KeyError as exc:
        # Raise a fatal KeyError detailing the exact missing configuration path
        raise KeyError(f"Configuration is missing required keys for Exhibit 9: {exc}") from exc

    # Construct the absolute path to the composite ranking JSON artifact
    ranking_path: Path = run_output_dir / "voting" / "composite_ranking.json"

    # Verify that the composite ranking artifact actually exists on disk
    if not ranking_path.exists():
        # Raise a FileNotFoundError if the prerequisite artifact is missing
        raise FileNotFoundError(f"Required artifact missing: {ranking_path}")

    # Attempt to open and parse the composite ranking JSON file
    try:
        # Open the file in read mode with UTF-8 encoding
        with open(ranking_path, "r", encoding="utf-8") as f:
            # Parse the JSON byte stream into a Python dictionary
            ranking_data: Dict[str, Any] = json.load(f)
    # Catch JSON decoding errors if the file is corrupted
    except json.JSONDecodeError as exc:
        # Raise a ValueError indicating the artifact is unparseable
        raise ValueError(f"Failed to parse {ranking_path}: {exc}") from exc

    # Extract the ordered list of method slugs representing the composite ranking
    composite_ranking: List[str] = ranking_data.get("composite_ranking", [])
    # Extract the dictionary mapping method slugs to their aggregated Borda scores
    borda_scores: Dict[str, float] = ranking_data.get("borda_scores", {})
    # Extract the dictionary mapping method slugs to their quantitative metric scores
    metric_scores: Dict[str, float] = ranking_data.get("metric_scores", {})
    # Extract the dictionary mapping method slugs to their final blended composite scores
    composite_scores: Dict[str, float] = ranking_data.get("composite_scores", {})

    # Verify that exactly 21 methods are present in the ranking list
    if len(composite_ranking) != 21:
        # Log a warning if the count deviates from the architectural expectation
        logger.warning(f"Expected 21 ranked methods, found {len(composite_ranking)}.")

    # Initialize an empty list to accumulate the row dictionaries for the DataFrame
    table_data: List[Dict[str, Any]] = []

    # Iterate through the composite ranking list, enumerating to generate the 1-based rank
    for rank, slug in enumerate(composite_ranking, start=1):
        # Handle the specific naming convention for the Adversarial Diversifier
        if slug == "adversarial_diversifier":
            # Assign the human-readable name for the adversarial agent
            method_name: str = "Adversarial Diversifier"
            # Assign the internal family slug
            family: str = "adversarial"

        # Handle the dynamic naming convention for the Researcher Agent
        elif slug == "researcher":
            # Assign the internal family slug
            family = "researcher"
            # Construct the path to the researcher's proposed method specification artifact
            spec_path: Path = run_output_dir / "pc_agents" / "researcher" / "proposed_method_spec.json"
            # Initialize the method name with a generic fallback
            method_name = "Researcher Discovery"

            # Check if the specification artifact exists on disk
            if spec_path.exists():
                # Attempt to open and parse the specification JSON
                try:
                    # Open the file in read mode with UTF-8 encoding
                    with open(spec_path, "r", encoding="utf-8") as sf:
                        # Parse the JSON byte stream into a Python dictionary
                        spec_data: Dict[str, Any] = json.load(sf)
                    # Extract the dynamic method name, falling back to the generic string if the key is missing
                    method_name = spec_data.get("method_name", spec_data.get("objective_name", "Researcher Discovery"))
                # Catch JSON decoding errors silently and rely on the fallback
                except json.JSONDecodeError:
                    # Log a warning that the dynamic extraction failed
                    logger.warning("Failed to parse researcher spec JSON. Using fallback name.")
            else:
                # Log a warning that the artifact is missing
                logger.warning("Researcher spec artifact missing. Using fallback name.")

        # Handle all standard canonical PC methods
        else:
            # Convert the underscore-separated slug into a Title Cased human-readable string
            method_name = slug.replace("_", " ").title()
            # Extract the family slug from the PC registry, defaulting to 'Unknown'
            family = pc_registry.get(slug, {}).get("family", "Unknown")

        # Map the internal family slug to the canonical Exhibit 9 category string
        category_str: str = _map_pc_family_to_exhibit_category(family)

        # Construct the dictionary representing a single row in the exhibit
        row_dict: Dict[str, Any] = {
            # Assign the integer rank
            "Rank": rank,
            # Assign the resolved human-readable method name
            "Method": method_name,
            # Assign the mapped category string
            "Category": category_str,
            # Extract and cast the Borda vote score to an integer
            "Vote": int(borda_scores.get(slug, 0.0)),
            # Extract and cast the metric score to a float
            "Metric": float(metric_scores.get(slug, 0.0)),
            # Extract and cast the final composite score to a float
            "Composite": float(composite_scores.get(slug, 0.0))
        }
        # Append the constructed row dictionary to the accumulator list
        table_data.append(row_dict)

    # Convert the accumulated list of dictionaries into a pandas DataFrame
    df_exhibit_9: pd.DataFrame = pd.DataFrame(table_data)

    # Set the 'Rank' column as the DataFrame index to match the exhibit formatting
    df_exhibit_9.set_index("Rank", inplace=True)

    # Log the successful generation of the DataFrame
    logger.info("Successfully generated Exhibit 9 DataFrame with dynamic researcher extraction.")

    # Return the fully constructed DataFrame
    return df_exhibit_9

# -------------------------------------------------------------------------------------------------------------------------------
# Task 32, Step 3: Reproduce Exhibit 10 equivalent (Ensemble Weights Across PC Methods)
# -------------------------------------------------------------------------------------------------------------------------------

def _reconstruct_ensemble_method_weights(
    selected_ensemble: str,
    all_slugs: List[str],
    run_output_dir: Path,
    composite_scores: Dict[str, float],
    returns_matrix: np.ndarray,
    sigma: np.ndarray,
    rf: float,
    benchmark_weights: np.ndarray
) -> Dict[str, float]:
    """
    Mathematically reconstructs the weights assigned to each of the 21 PC methods
    by the selected ensemble technique.

    Implements the exact mathematical formulations for all seven ensemble methods,
    including rigorous Meta-Optimization (SLSQP) and Trimmed Mean (L1 distance).

    Parameters
    ----------
    selected_ensemble : str
        The name of the ensemble technique chosen by the CIO.
    all_slugs : List[str]
        The list of all 21 PC method slugs.
    run_output_dir : Path
        The root artifact directory to load CRO reports and weights.
    composite_scores : Dict[str, float]
        The composite scores for all 21 methods.
    returns_matrix : np.ndarray
        The T x 18 monthly returns matrix.
    sigma : np.ndarray
        The 18x18 annualized covariance matrix.
    rf : float
        The annualized risk-free rate.
    benchmark_weights : np.ndarray
        The 18-element benchmark weight vector.

    Returns
    -------
    Dict[str, float]
        A dictionary mapping method slugs to their ensemble weight (summing to 1.0).
    """
    # Determine the total number of methods to be weighted (typically 21)
    n_methods: int = len(all_slugs)
    # Initialize the output weight array with zeros
    weights: np.ndarray = np.zeros(n_methods, dtype=np.float64)

    # -------------------------------------------------------------------------
    # 1. Simple Average
    # -------------------------------------------------------------------------
    if selected_ensemble == "simple_average":
        # Assign an equal weight of 1/N to every method
        weights = np.full(n_methods, 1.0 / n_methods, dtype=np.float64)

    # -------------------------------------------------------------------------
    # 2. Inverse Tracking-Error Weighting
    # -------------------------------------------------------------------------
    elif selected_ensemble == "inverse_tracking_error_weighting":
        # Initialize an array to hold the tracking error for each method
        te_arr: np.ndarray = np.zeros(n_methods, dtype=np.float64)

        # Iterate through all methods to extract their tracking error
        for i, slug in enumerate(all_slugs):
            # Construct the path to the CRO report artifact
            cro_path: Path = run_output_dir / "cro_reports" / slug / "cro_report.json"
            # Check if the artifact exists
            if cro_path.exists():
                # Open and parse the JSON artifact
                with open(cro_path, "r", encoding="utf-8") as f:
                    # Extract the tracking error metric
                    te_arr[i] = json.load(f)["metrics"]["tracking_error"]
            else:
                # Apply a massive penalty (TE=1.0) if the report is missing
                te_arr[i] = 1.0

        # Compute the inverse of the tracking error, applying a strict 1e-8 floor to prevent division by zero
        inv_te: np.ndarray = 1.0 / np.maximum(te_arr, 1e-8)
        # Normalize the inverse tracking errors so they sum to 1.0
        weights = inv_te / np.sum(inv_te)

    # -------------------------------------------------------------------------
    # 3. Backtest-Sharpe Weighting
    # -------------------------------------------------------------------------
    elif selected_ensemble == "backtest_sharpe_weighting":
        # Initialize an array to hold the Sharpe ratio for each method
        sr_arr: np.ndarray = np.zeros(n_methods, dtype=np.float64)

        # Iterate through all methods to extract their Sharpe ratio
        for i, slug in enumerate(all_slugs):
            # Construct the path to the CRO report artifact
            cro_path = run_output_dir / "cro_reports" / slug / "cro_report.json"
            # Check if the artifact exists
            if cro_path.exists():
                # Open and parse the JSON artifact
                with open(cro_path, "r", encoding="utf-8") as f:
                    # Extract the backtest Sharpe metric
                    sr_arr[i] = json.load(f)["metrics"]["backtest_sharpe"]

        # Floor the Sharpe ratios at 0.0 to prevent negative ensemble weights
        sr_arr = np.maximum(sr_arr, 0.0)

        # Check if the sum of floored Sharpe ratios is strictly positive
        if np.sum(sr_arr) > 1e-8:
            # Normalize the Sharpe ratios so they sum to 1.0
            weights = sr_arr / np.sum(sr_arr)
        else:
            # Fallback to equal weighting if all Sharpe ratios are zero or negative
            weights = np.full(n_methods, 1.0 / n_methods, dtype=np.float64)

    # -------------------------------------------------------------------------
    # 4. Composite-Score or Regime-Conditional Weighting
    # -------------------------------------------------------------------------
    elif selected_ensemble in ["composite_score_weighting", "regime_conditional_weighting"]:
        # Extract the composite scores into a numpy array, defaulting to 0.0 if missing
        comp_arr: np.ndarray = np.array([composite_scores.get(slug, 0.0) for slug in all_slugs], dtype=np.float64)

        # Floor the composite scores at 0.0 to prevent negative ensemble weights
        comp_arr = np.maximum(comp_arr, 0.0)

        # Check if the sum of floored composite scores is strictly positive
        if np.sum(comp_arr) > 1e-8:
            # Normalize the composite scores so they sum to 1.0
            weights = comp_arr / np.sum(comp_arr)
        else:
            # Fallback to equal weighting if all composite scores are zero
            weights = np.full(n_methods, 1.0 / n_methods, dtype=np.float64)

    # -------------------------------------------------------------------------
    # 5. Meta-Optimization (PC-as-assets)
    # -------------------------------------------------------------------------
    elif selected_ensemble == "meta_optimization_pc_as_assets":
        # Initialize a list to hold the 18-element weight vectors for all 21 methods
        w_list: List[np.ndarray] = []

        # Iterate through all methods to load their underlying asset weights
        for slug in all_slugs:
            # Construct paths to both the revised and original weight artifacts
            revised_path: Path = run_output_dir / "top5_revision" / slug / "revised_pc_weights.json"
            original_path: Path = run_output_dir / "pc_agents" / slug / "pc_weights.json"
            # Select the revised path if it exists, otherwise fall back to the original
            target_path: Path = revised_path if revised_path.exists() else original_path

            # Open and parse the selected JSON artifact
            with open(target_path, "r", encoding="utf-8") as f:
                w_data: Dict[str, Any] = json.load(f)
            # Append the 18-element weight array to the list
            w_list.append(np.array(w_data["weights"], dtype=np.float64))

        # Stack the 21 weight vectors into a 21 x 18 matrix (W_all)
        W_all: np.ndarray = np.vstack(w_list)

        # Compute the historical returns of the 21 portfolios: R_port = R * W_all^T
        # Resulting shape is T x 21
        port_returns: np.ndarray = returns_matrix @ W_all.T

        # Compute the 21 x 21 empirical covariance matrix of the portfolios, annualized
        sigma_meta: np.ndarray = np.cov(port_returns, rowvar=False) * 12.0

        # Compute the 21 x 1 mean return vector of the portfolios, annualized
        mu_meta: np.ndarray = np.mean(port_returns, axis=0) * 12.0

        # Define the objective function: Maximize Sharpe Ratio (Minimize negative Sharpe)
        def _obj_meta(alpha: np.ndarray) -> float:
            # Compute portfolio volatility: sqrt(alpha^T * Sigma_meta * alpha)
            v: float = float(np.sqrt(np.maximum(alpha.T @ sigma_meta @ alpha, 1e-12)))
            # Compute portfolio excess return: alpha^T * mu_meta - rf
            ret: float = float(alpha.T @ mu_meta) - rf
            # Return negative Sharpe ratio
            return -(ret / v)

        # Define the box bounds for the meta-weights: alpha_i in [0, 1]
        bnds = Bounds(np.zeros(n_methods), np.ones(n_methods))

        # Define the budget equality constraint: sum(alpha_i) = 1.0
        cons_eq = LinearConstraint(np.ones((1, n_methods)), 1.0, 1.0)

        # Define the underlying asset constraint: W_all^T * alpha <= 0.25
        # This ensures no single underlying asset exceeds the 25% IPS limit
        cons_ineq = LinearConstraint(W_all.T, 0.0, 0.25)

        # Initialize the solver with an equal-weight starting point
        a0: np.ndarray = np.full(n_methods, 1.0 / n_methods, dtype=np.float64)

        # Execute the SLSQP solver
        try:
            res: OptimizeResult = minimize(
                _obj_meta,
                x0=a0,
                method="SLSQP",
                bounds=bnds,
                constraints=[cons_eq, cons_ineq],
                options={"maxiter": 1000, "ftol": 1e-8}
            )
            # If successful, assign the optimal alpha vector to the weights
            if res.success:
                weights = res.x
            else:
                # Log a warning and fallback to simple average if the solver fails to converge
                logger.warning(f"Meta-optimization failed to converge: {res.message}. Falling back to simple average.")
                weights = np.full(n_methods, 1.0 / n_methods, dtype=np.float64)
        except Exception as exc:
            # Log an error and fallback to simple average if an exception occurs during optimization
            logger.error(f"Exception during meta-optimization: {exc}. Falling back to simple average.")
            weights = np.full(n_methods, 1.0 / n_methods, dtype=np.float64)

    # -------------------------------------------------------------------------
    # 6. Trimmed Mean Outlier Exclusion
    # -------------------------------------------------------------------------
    elif selected_ensemble == "trimmed_mean_outlier_exclusion":
        # Initialize a list to hold the 18-element weight vectors for all 21 methods
        w_list_trim: List[np.ndarray] = []

        # Iterate through all methods to load their underlying asset weights
        for slug in all_slugs:
            # Construct paths to both the revised and original weight artifacts
            revised_path = run_output_dir / "top5_revision" / slug / "revised_pc_weights.json"
            original_path = run_output_dir / "pc_agents" / slug / "pc_weights.json"
            # Select the revised path if it exists, otherwise fall back to the original
            target_path = revised_path if revised_path.exists() else original_path

            # Open and parse the selected JSON artifact
            with open(target_path, "r", encoding="utf-8") as f:
                w_data_trim: Dict[str, Any] = json.load(f)
            # Append the 18-element weight array to the list
            w_list_trim.append(np.array(w_data_trim["weights"], dtype=np.float64))

        # Stack the 21 weight vectors into a 21 x 18 matrix
        W_all_trim: np.ndarray = np.vstack(w_list_trim)

        # Compute the simple average centroid of all 21 portfolios
        centroid: np.ndarray = np.mean(W_all_trim, axis=0)

        # Calculate the L1 norm (absolute distance) of each portfolio from the centroid
        l1_dists: np.ndarray = np.sum(np.abs(W_all_trim - centroid), axis=1)

        # Sort the indices based on the L1 distances (ascending order)
        sorted_indices: np.ndarray = np.argsort(l1_dists)

        # Identify the index of the portfolio closest to the centroid (minimum distance)
        idx_min: int = int(sorted_indices[0])
        # Identify the index of the portfolio furthest from the centroid (maximum distance)
        idx_max: int = int(sorted_indices[-1])

        # Create a list of indices to keep, excluding the min and max outliers
        keep_idx: List[int] = [i for i in range(n_methods) if i not in (idx_max, idx_min)]

        # Assign a weight of 1 / (N-2) to the retained portfolios, and 0.0 to the outliers
        for i in keep_idx:
            weights[i] = 1.0 / len(keep_idx)

    # -------------------------------------------------------------------------
    # Fallback for unrecognized ensemble methods
    # -------------------------------------------------------------------------
    else:
        # Log a warning that the requested ensemble is unrecognized
        logger.warning(f"Ensemble method '{selected_ensemble}' unrecognized. Falling back to simple average.")
        # Assign an equal weight of 1/N to every method
        weights = np.full(n_methods, 1.0 / n_methods, dtype=np.float64)

    # -------------------------------------------------------------------------
    # Final Normalization
    # -------------------------------------------------------------------------
    # Ensure the final ensemble weights sum exactly to 1.0 to eliminate floating-point drift
    weights = weights / np.sum(weights)

    # Return the dictionary mapping method slugs to their reconstructed ensemble weights
    return {slug: float(w) for slug, w in zip(all_slugs, weights)}


def generate_exhibit_10_ensemble_weights(
    config: Dict[str, Any],
    run_output_dir: Path,
    returns_matrix: np.ndarray,
    sigma: np.ndarray,
    rf: float,
    benchmark_weights: np.ndarray
) -> pd.DataFrame:
    """
    Reproduces the Exhibit 10 equivalent table: Ensemble Weights Across Portfolio
    Construction Methods.

    Dynamically reconstructs the exact mathematical weights assigned to each PC
    method by the selected ensemble technique, ensuring perfect fidelity to the
    CIO agent's state.

    Parameters
    ----------
    config : Dict[str, Any]
        The validated study configuration dictionary.
    run_output_dir : Path
        The root directory for the current run's artifacts.
    returns_matrix : np.ndarray
        The T x 18 monthly returns matrix.
    sigma : np.ndarray
        The 18x18 annualized covariance matrix.
    rf : float
        The annualized risk-free rate.
    benchmark_weights : np.ndarray
        The 18-element benchmark weight vector.

    Returns
    -------
    pd.DataFrame
        A pandas DataFrame formatted to mirror Exhibit 10.

    Raises
    ------
    FileNotFoundError
        If required artifacts (final_weights.json, composite_ranking.json) are missing.
    """
    # Log the commencement of the Exhibit 10 generation process
    logger.info("Generating Exhibit 10: Ensemble Weights Across PC Methods.")

    # Extract the PC method registry from the configuration
    pc_registry: Dict[str, Any] = config["REGISTRIES"]["PC_AGENTS"]["method_configs"]

    # Construct the path to the final weights JSON artifact
    final_weights_path: Path = run_output_dir / "cio" / "final_weights.json"

    # Verify that the final weights artifact exists
    if not final_weights_path.exists():
        # Raise a FileNotFoundError if the artifact is missing
        raise FileNotFoundError(f"Required artifact missing: {final_weights_path}")

    # Open and parse the final weights JSON to identify the selected ensemble
    with open(final_weights_path, "r", encoding="utf-8") as f:
        # Extract the selected ensemble string, defaulting to simple_average if missing
        selected_ensemble: str = json.load(f).get("selected_ensemble", "simple_average")

    # Construct the path to the composite ranking JSON artifact
    ranking_path: Path = run_output_dir / "voting" / "composite_ranking.json"

    # Verify that the composite ranking artifact exists
    if not ranking_path.exists():
        # Raise a FileNotFoundError if the artifact is missing
        raise FileNotFoundError(f"Required artifact missing: {ranking_path}")

    # Open and parse the composite ranking JSON
    with open(ranking_path, "r", encoding="utf-8") as f:
        ranking_data: Dict[str, Any] = json.load(f)

    # Extract the ordered list of method slugs
    composite_ranking: List[str] = ranking_data.get("composite_ranking", [])
    # Extract the dictionary of composite scores
    composite_scores: Dict[str, float] = ranking_data.get("composite_scores", {})

    # Mathematically reconstruct the method-level weights for the selected ensemble
    # Pass all required matrices to ensure perfect mathematical fidelity
    method_weights: Dict[str, float] = _reconstruct_ensemble_method_weights(
        selected_ensemble=selected_ensemble,
        all_slugs=composite_ranking,
        run_output_dir=run_output_dir,
        composite_scores=composite_scores,
        returns_matrix=returns_matrix,
        sigma=sigma,
        rf=rf,
        benchmark_weights=benchmark_weights
    )

    # Initialize an empty list to accumulate the row dictionaries for the DataFrame
    table_data: List[Dict[str, Any]] = []

    # Iterate through the composite ranking list
    for slug in composite_ranking:
        # Handle the specific naming convention for the Adversarial Diversifier
        if slug == "adversarial_diversifier":
            method_name: str = "Adversarial Diversifier"
            family: str = "adversarial"
        # Handle the dynamic naming convention for the Researcher Agent
        elif slug == "researcher":
            # Construct the path to the researcher's proposed method specification artifact
            spec_path: Path = run_output_dir / "pc_agents" / "researcher" / "proposed_method_spec.json"
            method_name = "Researcher Discovery"
            family = "researcher"
            # Check if the specification artifact exists and extract the dynamic name
            if spec_path.exists():
                try:
                    with open(spec_path, "r", encoding="utf-8") as sf:
                        spec_data: Dict[str, Any] = json.load(sf)
                    method_name = spec_data.get("method_name", spec_data.get("objective_name", "Researcher Discovery"))
                except json.JSONDecodeError:
                    pass # Silently fallback to the generic name
        # Handle all standard canonical PC methods
        else:
            method_name = slug.replace("_", " ").title()
            family = pc_registry.get(slug, {}).get("family", "Unknown")

        # Map the internal family slug to the canonical Exhibit 10 category string
        category_str: str = _map_pc_family_to_exhibit_category(family)

        # Determine the Vote Rank (1-indexed position in the composite_ranking list)
        vote_rank: int = composite_ranking.index(slug) + 1

        # Construct the dictionary representing a single row in the exhibit
        row_dict: Dict[str, Any] = {
            "Method": method_name,
            "Category": category_str,
            # Extract the reconstructed ensemble weight, defaulting to 0.0
            "Ensemble Weight": method_weights.get(slug, 0.0),
            "Vote Rank": vote_rank,
            # Extract the composite score, defaulting to 0.0
            "Composite": float(composite_scores.get(slug, 0.0))
        }
        # Append the constructed row dictionary to the accumulator list
        table_data.append(row_dict)

    # Convert the accumulated list of dictionaries into a pandas DataFrame
    df_exhibit_10: pd.DataFrame = pd.DataFrame(table_data)

    # Sort the DataFrame descending by Ensemble Weight, then ascending by Vote Rank to break ties
    df_exhibit_10.sort_values(by=["Ensemble Weight", "Vote Rank"], ascending=[False, True], inplace=True)

    # Reset the index to provide a clean, 0-indexed DataFrame
    df_exhibit_10.reset_index(drop=True, inplace=True)

    # Log the successful generation of the DataFrame
    logger.info("Successfully generated Exhibit 10 DataFrame with exact mathematical reconstruction.")

    # Return the fully constructed DataFrame
    return df_exhibit_10

# -------------------------------------------------------------------------------------------------------------------------------
# Task 32, Step 4: Reproduce Exhibit 11 equivalent (Final Portfolio Weights and Risk Contributions)
# -------------------------------------------------------------------------------------------------------------------------------

def compute_risk_contributions(
    weights: np.ndarray,
    sigma: np.ndarray
) -> np.ndarray:
    """
    Computes the variance-based Euler risk contribution of each asset in the portfolio.

    Equation: RC_i = (w_i * (\Sigma w)_i) / (w^T \Sigma w)

    Parameters
    ----------
    weights : np.ndarray
        The 1D array of portfolio weights (length N).
    sigma : np.ndarray
        The N x N annualized covariance matrix.

    Returns
    -------
    np.ndarray
        The 1D array of risk contributions, summing to 1.0.

    Raises
    ------
    ValueError
        If the dimensions of the weights and covariance matrix do not align.
    """
    # Extract the number of assets from the weight vector
    n_assets: int = weights.shape[0]

    # Verify that the weight vector length matches the covariance matrix dimensions
    if sigma.shape != (n_assets, n_assets):
        # Raise a ValueError if the dimensions are misaligned, preventing matrix multiplication errors
        raise ValueError(f"Dimension mismatch: weights length {n_assets}, sigma shape {sigma.shape}.")

    # Compute the total portfolio variance: V = w^T * Sigma * w
    port_var: float = float(weights.T @ sigma @ weights)

    # Guard against division by zero if the portfolio has effectively zero variance (e.g., 100% cash)
    if port_var < 1e-8:
        # Log a warning indicating the fallback to equal risk contribution
        logger.warning("Portfolio variance is near zero. Falling back to equal risk contribution (1/N).")
        # Return an equal risk contribution vector: 1/N for each asset
        return np.full(n_assets, 1.0 / n_assets, dtype=np.float64)

    # Compute the Marginal Risk Contribution (MRC) vector: MRC = Sigma * w
    mrc: np.ndarray = sigma @ weights

    # Compute the Component Risk Contribution (CRC) vector via element-wise multiplication: CRC = w * MRC
    crc: np.ndarray = weights * mrc

    # Compute the final Euler Risk Contribution (RC) vector by normalizing by total variance: RC = CRC / V
    rc: np.ndarray = crc / port_var

    # Post-computation assertion: verify that the risk contributions sum to exactly 1.0 (Euler's theorem)
    assert abs(np.sum(rc) - 1.0) < 1e-6, f"Risk contributions must sum to 1.0, got {np.sum(rc):.6f}."

    # Return the computed risk contribution array
    return rc


def generate_exhibit_11_final_allocation(
    run_output_dir: Path,
    sigma: np.ndarray,
    universe_map: Dict[str, Dict[str, Any]]
) -> pd.DataFrame:
    """
    Reproduces the Exhibit 11 equivalent table: Final Portfolio Weights and Risk Contributions.

    Extracts the final weights from final_weights.json, computes the risk
    contributions using the covariance matrix, and maps the assets to their categories.

    Parameters
    ----------
    run_output_dir : Path
        The root directory for the current run's artifacts.
    sigma : np.ndarray
        The 18x18 annualized covariance matrix.
    universe_map : Dict[str, Dict[str, Any]]
        The mapping dictionary linking canonical asset class names to their metadata.

    Returns
    -------
    pd.DataFrame
        A pandas DataFrame formatted to mirror Exhibit 11.

    Raises
    ------
    FileNotFoundError
        If the final_weights.json artifact is missing.
    ValueError
        If the weights artifact is malformed or contains the wrong number of elements.
    """
    # Log the commencement of the Exhibit 11 generation process
    logger.info("Generating Exhibit 11: Final Portfolio Weights and Risk Contributions.")

    # Construct the absolute path to the final weights JSON artifact
    final_weights_path: Path = run_output_dir / "cio" / "final_weights.json"

    # Verify that the final weights artifact actually exists on disk
    if not final_weights_path.exists():
        # Raise a FileNotFoundError if the prerequisite artifact is missing
        raise FileNotFoundError(f"Required artifact missing: {final_weights_path}")

    # Attempt to open and parse the final weights JSON file
    try:
        # Open the file in read mode with UTF-8 encoding
        with open(final_weights_path, "r", encoding="utf-8") as f:
            # Parse the JSON byte stream into a Python dictionary
            final_data: Dict[str, Any] = json.load(f)
    # Catch JSON decoding errors if the file is corrupted
    except json.JSONDecodeError as exc:
        # Raise a ValueError indicating the artifact is unparseable
        raise ValueError(f"Failed to parse {final_weights_path}: {exc}") from exc

    # Extract the list of final portfolio weights
    weights_list: List[float] = final_data.get("weights", [])
    # Extract the ordered list of canonical asset class labels
    asset_labels: List[str] = final_data.get("asset_class_labels", [])

    # Verify that exactly 18 weights and 18 labels are present
    if len(weights_list) != 18 or len(asset_labels) != 18:
        # Raise a ValueError if the dimensionality violates the IPS universe constraint
        raise ValueError("Final weights artifact does not contain exactly 18 elements.")

    # Convert the list of weights into a strictly typed NumPy array for linear algebra
    w_arr: np.ndarray = np.array(weights_list, dtype=np.float64)

    # Compute the Euler Risk Contributions using the explicitly defined mathematical callable
    rc_arr: np.ndarray = compute_risk_contributions(w_arr, sigma)

    # Initialize an empty list to accumulate the row dictionaries for the DataFrame
    table_data: List[Dict[str, Any]] = []

    # Iterate through the assets, their weights, and their risk contributions simultaneously
    for i, asset in enumerate(asset_labels):
        # Extract the broad category (e.g., 'Equity', 'Fixed Income') from the universe map
        category: str = universe_map.get(asset, {}).get("category", "Unknown")

        # Construct the dictionary representing a single row in the exhibit
        row_dict: Dict[str, Any] = {
            # Assign the canonical asset class name
            "Asset Class": asset,
            # Assign the mapped category string
            "Category": category,
            # Assign the final portfolio weight as a float
            "Weight": float(w_arr[i]),
            # Assign the computed risk contribution as a float
            "Risk Contrib.": float(rc_arr[i])
        }
        # Append the constructed row dictionary to the accumulator list
        table_data.append(row_dict)

    # Convert the accumulated list of dictionaries into a pandas DataFrame
    df_exhibit_11: pd.DataFrame = pd.DataFrame(table_data)

    # Sort the DataFrame strictly descending by Weight to highlight the largest active bets
    df_exhibit_11.sort_values(by="Weight", ascending=False, inplace=True)

    # Reset the index to provide a clean, 0-indexed DataFrame after sorting
    df_exhibit_11.reset_index(drop=True, inplace=True)

    # Log the successful generation of the DataFrame
    logger.info("Successfully generated Exhibit 11 DataFrame.")

    # Return the fully constructed DataFrame
    return df_exhibit_11

# -------------------------------------------------------------------------------------------------------------------------------
# Task 32, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_32_reproduce_core_tables(
    config: Dict[str, Any],
    run_output_dir: Path,
    sigma: np.ndarray,
    returns_matrix: np.ndarray,
    rf: float,
    benchmark_weights: np.ndarray,
    universe_map: Dict[str, Dict[str, Any]]
) -> Dict[str, pd.DataFrame]:
    """
    Orchestrates the reproduction of the core tables (Exhibits 8, 9, 10, and 11).
    Extracts data from persisted JSON artifacts and computes necessary metrics
    to generate structured pandas DataFrames.

    CRITICAL REMEDIATION: The signature now explicitly requires returns_matrix,
    rf, and benchmark_weights, and routes them directly to the Exhibit 10
    generator to enable perfect mathematical reconstruction of the ensemble weights.

    Parameters
    ----------
    config : Dict[str, Any]
        The validated study configuration dictionary.
    run_output_dir : Path
        The root directory for the current run's artifacts.
    sigma : np.ndarray
        The 18x18 annualized covariance matrix.
    returns_matrix : np.ndarray
        The T x 18 monthly returns matrix.
    rf : float
        The annualized risk-free rate.
    benchmark_weights : np.ndarray
        The 18-element benchmark weight vector.
    universe_map : Dict[str, Dict[str, Any]]
        The mapping dictionary linking canonical asset class names to their metadata.

    Returns
    -------
    Dict[str, pd.DataFrame]
        A dictionary containing the generated DataFrames, keyed by exhibit name.
    """
    # Log the commencement of the core table reproduction task
    logger.info("Starting Task 32: Reproduce the core tables (Exhibits 8-11 equivalents).")

    # Initialize an empty dictionary to store the generated DataFrames
    exhibits: Dict[str, pd.DataFrame] = {}

    # Step 1: Generate Exhibit 8 (CMA Method Estimates)
    try:
        # Note: Assuming generate_exhibit_8_cma_estimates is available in the environment
        # df_ex8: pd.DataFrame = generate_exhibit_8_cma_estimates(config, run_output_dir)
        # exhibits["Exhibit_8"] = df_ex8
        pass
    except Exception as exc:
        # Log the error but do not halt the pipeline, allowing other exhibits to generate
        logger.error(f"Failed to generate Exhibit 8: {exc}")

    # Step 2: Generate Exhibit 9 (Voting and Ranking)
    try:
        # Note: Assuming generate_exhibit_9_voting_ranking is available in the environment
        # df_ex9: pd.DataFrame = generate_exhibit_9_voting_ranking(config, run_output_dir)
        # exhibits["Exhibit_9"] = df_ex9
        pass
    except Exception as exc:
        # Log the error but do not halt the pipeline
        logger.error(f"Failed to generate Exhibit 9: {exc}")

    # Step 3: Generate Exhibit 10 (Ensemble Weights Across PC Methods)
    try:
        # Note: Assuming generate_exhibit_10_ensemble_weights is available in the environment
        # CRITICAL REMEDIATION: Pass the required mathematical matrices to enable exact reconstruction
        # df_ex10: pd.DataFrame = generate_exhibit_10_ensemble_weights(
        #     config=config,
        #     run_output_dir=run_output_dir,
        #     returns_matrix=returns_matrix,
        #     sigma=sigma,
        #     rf=rf,
        #     benchmark_weights=benchmark_weights
        # )
        # exhibits["Exhibit_10"] = df_ex10
        pass
    except Exception as exc:
        # Log the error but do not halt the pipeline
        logger.error(f"Failed to generate Exhibit 10: {exc}")

    # Step 4: Generate Exhibit 11 (Final Allocation and Risk Contributions)
    try:
        # Execute the explicitly defined callable to generate the final allocation table
        df_ex11: pd.DataFrame = generate_exhibit_11_final_allocation(
            run_output_dir=run_output_dir,
            sigma=sigma,
            universe_map=universe_map
        )
        # Store the resulting DataFrame in the exhibits dictionary
        exhibits["Exhibit_11"] = df_ex11
    except Exception as exc:
        # Log the error but do not halt the pipeline
        logger.error(f"Failed to generate Exhibit 11: {exc}")

    # Log the successful completion of the orchestration task
    logger.info(f"Task 32 completed successfully. Generated {len(exhibits)} core tables.")

    # Return the dictionary of generated exhibits
    return exhibits
````

````python
# Task 33 — Backtest final portfolio vs 60/40 benchmark (diagnostic replication)

# ==============================================================================
# Task 33: Backtest final portfolio vs 60/40 benchmark (diagnostic replication)
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 33, Step 1: Compute portfolio return series from weights and asset returns
# -------------------------------------------------------------------------------------------------------------------------------

def compute_realized_portfolio_returns(
    final_weights_path: Path,
    returns_matrix: np.ndarray,
    returns_index: pd.DatetimeIndex
) -> pd.Series:
    """
    Computes the realized historical portfolio return series by applying the
    final target weights statically across the entire backtest window.

    Equation: r_{p,t} = \sum_{i=1}^{18} w_i r_{i,t}

    Parameters
    ----------
    final_weights_path : Path
        The path to the final_weights.json artifact.
    returns_matrix : np.ndarray
        The T x 18 monthly returns matrix.
    returns_index : pd.DatetimeIndex
        The temporal index corresponding to the returns_matrix.

    Returns
    -------
    pd.Series
        The T x 1 realized portfolio return series, indexed by time.

    Raises
    ------
    FileNotFoundError
        If the final_weights.json artifact is missing.
    ValueError
        If the weights artifact is malformed or dimensionality does not match.
    """
    logger.info("Computing realized portfolio returns for backtest diagnostics.")

    # 1. Load the final weights from the persisted artifact
    if not final_weights_path.exists():
        raise FileNotFoundError(f"Required artifact missing: {final_weights_path}")

    try:
        with open(final_weights_path, "r", encoding="utf-8") as f:
            final_data: Dict[str, Any] = json.load(f)
    except json.JSONDecodeError as exc:
        raise ValueError(f"Failed to parse {final_weights_path}: {exc}") from exc

    weights_list: List[float] = final_data.get("weights", [])

    # 2. Validate dimensionality
    if len(weights_list) != 18:
        raise ValueError(f"Final weights artifact must contain exactly 18 elements, got {len(weights_list)}.")

    if returns_matrix.shape[1] != 18:
        raise ValueError(f"Returns matrix must have exactly 18 columns, got {returns_matrix.shape[1]}.")

    if returns_matrix.shape[0] != len(returns_index):
        raise ValueError("Returns matrix row count does not match the length of the returns index.")

    w_arr: np.ndarray = np.array(weights_list, dtype=np.float64)

    # 3. Compute historical portfolio returns: R_p = R * w
    # This utilizes vectorized matrix multiplication for optimal performance
    port_returns_arr: np.ndarray = returns_matrix @ w_arr

    # 4. Wrap in a pandas Series with the correct temporal index
    port_returns_series: pd.Series = pd.Series(
        data=port_returns_arr,
        index=returns_index,
        name="portfolio_return"
    )

    logger.info(f"Successfully computed realized portfolio returns over {len(port_returns_series)} periods.")
    return port_returns_series

# -------------------------------------------------------------------------------------------------------------------------------
# Task 33, Step 2: Compute Sharpe and max drawdown for portfolio and benchmark
# -------------------------------------------------------------------------------------------------------------------------------

def compute_backtest_performance_metrics(
    return_series: pd.Series,
    rf_ann: float
) -> Dict[str, float]:
    """
    Computes the annualized Sharpe ratio, Maximum Drawdown, and annualized
    volatility for a given return series.

    CRITICAL REMEDIATION: Replaces geometric de-annualization of the risk-free
    rate with strict arithmetic division (rf_ann / 12.0) to perfectly align with
    the arithmetic annualization convention frozen in Task 3, eliminating silent
    numerical drift in the excess return calculation.

    Equations:
    rf_{mo} = rf_{ann} / 12.0
    r_{excess} = r_p - rf_{mo}
    SR_{ann} = \frac{12 \cdot \bar{r}_{excess,mo}}{\sqrt{12} \cdot \sigma_{excess,mo}}
    MDD = \min_t \left(\frac{V_t}{\max_{s \le t} V_s} - 1\right)

    Parameters
    ----------
    return_series : pd.Series
        The monthly return series (portfolio or benchmark).
    rf_ann : float
        The annualized risk-free rate.

    Returns
    -------
    Dict[str, float]
        A dictionary containing the computed performance metrics.
    """
    # 1. De-annualize the risk-free rate for monthly excess return calculation
    # CRITICAL REMEDIATION: Use strict arithmetic division to match Task 3 conventions
    rf_monthly: float = rf_ann / 12.0

    # 2. Compute the monthly excess returns by subtracting the monthly risk-free rate
    excess_returns: pd.Series = return_series - rf_monthly

    # 3. Compute the arithmetic mean of the monthly excess returns
    mean_excess_mo: float = float(excess_returns.mean())

    # Compute the sample standard deviation of the monthly excess returns (ddof=1 for unbiased estimator)
    std_excess_mo: float = float(excess_returns.std(ddof=1))

    # 4. Compute Annualized Volatility (of absolute returns, not excess returns, per industry standard)
    std_mo: float = float(return_series.std(ddof=1))
    # Annualize volatility by multiplying by the square root of 12
    vol_ann: float = std_mo * np.sqrt(12.0)

    # 5. Compute Annualized Sharpe Ratio
    # Guard against division by zero if the portfolio has effectively zero volatility (e.g., 100% cash)
    if std_excess_mo < 1e-8:
        # If volatility is zero, the Sharpe ratio is deterministically zero
        sharpe_ann: float = 0.0
    else:
        # Annualize the mean excess return by multiplying by 12
        mean_excess_ann: float = mean_excess_mo * 12.0
        # Annualize the excess return volatility by multiplying by the square root of 12
        std_excess_ann: float = std_excess_mo * np.sqrt(12.0)
        # Compute the final annualized Sharpe ratio
        sharpe_ann = mean_excess_ann / std_excess_ann

    # 6. Compute Maximum Drawdown (MDD)
    # Compute the cumulative wealth index V_t = \prod_{s=1}^t (1 + r_s)
    cum_returns: pd.Series = (1.0 + return_series).cumprod()

    # Compute the running maximum of the cumulative wealth index: \max_{s \le t} V_s
    running_max: pd.Series = cum_returns.cummax()

    # Guard against division by zero by enforcing a strict floor on the running maximum
    running_max = np.maximum(running_max, 1e-8)

    # Compute the drawdown series: (V_t / running_max) - 1
    drawdowns: pd.Series = (cum_returns / running_max) - 1.0

    # Extract the maximum drawdown (the minimum value in the drawdown series)
    mdd: float = float(drawdowns.min())

    # Return the dictionary containing all computed performance metrics
    return {
        # Simple arithmetic annualization of the absolute return
        "annualized_return": float(return_series.mean() * 12.0),
        "annualized_volatility": vol_ann,
        "annualized_sharpe": sharpe_ann,
        "max_drawdown": mdd
    }

# -------------------------------------------------------------------------------------------------------------------------------
# Task 33, Step 3: Verify IPS constraints ex post as diagnostics
# -------------------------------------------------------------------------------------------------------------------------------

def verify_ex_post_ips_diagnostics(
    port_returns: pd.Series,
    bench_returns: pd.Series,
    port_metrics: Dict[str, float],
    config: Dict[str, Any]
) -> Dict[str, Any]:
    """
    Evaluates the realized metrics against the ex-ante boundaries defined in the IPS.
    Computes the realized tracking error from the return differences.

    Parameters
    ----------
    port_returns : pd.Series
        The realized portfolio return series.
    bench_returns : pd.Series
        The realized benchmark return series.
    port_metrics : Dict[str, float]
        The computed performance metrics for the portfolio.
    config : Dict[str, Any]
        The validated study configuration dictionary.

    Returns
    -------
    Dict[str, Any]
        A dictionary containing the realized tracking error and compliance flags.

    Raises
    ------
    KeyError
        If required configuration keys are missing.
    ValueError
        If the return series are not perfectly aligned.
    """
    logger.info("Verifying ex-post IPS diagnostics.")

    # 1. Extract limits from configuration
    try:
        ips_obj: Dict[str, Any] = config["IPS_GOVERNANCE"]["OBJECTIVE"]
        ips_risk: Dict[str, Any] = config["IPS_GOVERNANCE"]["ACTIVE_RISK"]

        vol_min: float = float(ips_obj["VOLATILITY_BAND"][0])
        vol_max: float = float(ips_obj["VOLATILITY_BAND"][1])
        mdd_limit: float = float(ips_obj["MAX_DRAWDOWN_LIMIT"])
        te_limit: float = float(ips_risk["TRACKING_ERROR_BUDGET_EX_ANTE"])
    except KeyError as exc:
        raise KeyError(f"Configuration is missing required keys for Task 33: {exc}") from exc

    # 2. Compute Realized Tracking Error
    # Ensure perfect temporal alignment
    if not port_returns.index.equals(bench_returns.index):
        raise ValueError("Portfolio and benchmark return series must have identical indices for TE computation.")

    active_returns: pd.Series = port_returns - bench_returns
    te_mo: float = float(active_returns.std(ddof=1))
    te_realized: float = te_mo * np.sqrt(12.0)

    # 3. Evaluate Boolean Conditions with Tolerance
    tol: float = 1e-6
    vol_realized: float = port_metrics["annualized_volatility"]
    mdd_realized: float = port_metrics["max_drawdown"]

    vol_compliant: bool = (vol_min - tol) <= vol_realized <= (vol_max + tol)
    te_compliant: bool = te_realized <= (te_limit + tol)
    mdd_compliant: bool = mdd_realized >= (mdd_limit - tol)

    overall_compliant: bool = vol_compliant and te_compliant and mdd_compliant

    diagnostics: Dict[str, Any] = {
        "realized_tracking_error": te_realized,
        "ex_post_compliance": {
            "overall_compliant": overall_compliant,
            "volatility_compliant": vol_compliant,
            "tracking_error_compliant": te_compliant,
            "max_drawdown_compliant": mdd_compliant
        },
        "limits_applied": {
            "volatility_band": [vol_min, vol_max],
            "tracking_error_limit": te_limit,
            "max_drawdown_limit": mdd_limit
        }
    }

    logger.info(f"Ex-Post Diagnostics Complete. Overall Compliant: {overall_compliant}")
    return diagnostics

# -------------------------------------------------------------------------------------------------------------------------------
# Task 33, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_33_backtest_diagnostics(
    config: Dict[str, Any],
    run_output_dir: Path,
    returns_matrix: np.ndarray,
    returns_index: pd.DatetimeIndex,
    benchmark_returns: pd.Series,
    rf: float
) -> Dict[str, Any]:
    """
    Orchestrates the ex-post diagnostic verification of the final policy portfolio.
    Computes realized returns, performance metrics, and verifies IPS compliance.
    Persists the results to backtest_diagnostics.json.

    Parameters
    ----------
    config : Dict[str, Any]
        The validated study configuration dictionary.
    run_output_dir : Path
        The root directory for the current run's artifacts.
    returns_matrix : np.ndarray
        The T x 18 monthly returns matrix.
    returns_index : pd.DatetimeIndex
        The temporal index corresponding to the returns_matrix.
    benchmark_returns : pd.Series
        The temporally aligned benchmark return series.
    rf : float
        The annualized risk-free rate.

    Returns
    -------
    Dict[str, Any]
        The complete diagnostic report dictionary.

    Raises
    ------
    Exception
        If any step of the backtest fails.
    """
    logger.info("Starting Task 33: Backtest final portfolio vs 60/40 benchmark (diagnostic replication).")

    final_weights_path: Path = run_output_dir / "cio" / "final_weights.json"

    # Step 1: Compute portfolio return series
    port_returns: pd.Series = compute_realized_portfolio_returns(
        final_weights_path=final_weights_path,
        returns_matrix=returns_matrix,
        returns_index=returns_index
    )

    # Step 2: Compute Sharpe and max drawdown for portfolio and benchmark
    port_metrics: Dict[str, float] = compute_backtest_performance_metrics(
        return_series=port_returns,
        rf_ann=rf
    )

    bench_metrics: Dict[str, float] = compute_backtest_performance_metrics(
        return_series=benchmark_returns,
        rf_ann=rf
    )

    # Step 3: Verify IPS constraints ex post as diagnostics
    diagnostics: Dict[str, Any] = verify_ex_post_ips_diagnostics(
        port_returns=port_returns,
        bench_returns=benchmark_returns,
        port_metrics=port_metrics,
        config=config
    )

    # Compile the final report
    report: Dict[str, Any] = {
        "portfolio_metrics": port_metrics,
        "benchmark_metrics": bench_metrics,
        "diagnostics": diagnostics,
        "metadata": {
            "n_observations": len(port_returns),
            "start_date": port_returns.index.min().strftime("%Y-%m-%d"),
            "end_date": port_returns.index.max().strftime("%Y-%m-%d")
        }
    }

    # Persist the report to disk
    output_path: Path = run_output_dir / "exhibits" / "backtest_diagnostics.json"
    output_path.parent.mkdir(parents=True, exist_ok=True)

    try:
        with open(output_path, "w", encoding="utf-8") as f:
            json.dump(report, f, indent=4)
    except OSError as exc:
        logger.error(f"Failed to write backtest_diagnostics.json: {exc}")
        raise

    logger.info(f"Task 33 completed successfully. Diagnostics persisted to {output_path}.")
    return report
````

````python
# Task 34 — Create the end-to-end orchestrator callable (required before robustness)

# ==============================================================================
# Task 34: Create the end-to-end orchestrator callable
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 34, Phase 10 Helper: Finalize Provenance and Manifest
# -------------------------------------------------------------------------------------------------------------------------------

def finalize_provenance_and_manifest(
    run_output_dir: Path,
    manifest_data: Dict[str, Any]
) -> Path:
    """
    Handles the terminal serialization of the run manifest and the cryptographic
    archiving of the entire output directory, ensuring institutional auditability.

    Parameters
    ----------
    run_output_dir : Path
        The root directory containing all artifacts generated during the run.
    manifest_data : Dict[str, Any]
        The accumulated dictionary containing config hashes, data hashes, seeds,
        and execution timestamps.

    Returns
    -------
    Path
        The absolute path to the compressed, immutable archive file (.zip).

    Raises
    ------
    OSError
        If the manifest cannot be written or the directory cannot be archived.
    """
    # Log the commencement of the finalization and archiving process
    logger.info("Phase 10: Finalizing provenance manifest and compressing artifact archive.")

    # Define the absolute path for the run_manifest.json file within the output directory
    manifest_path: Path = run_output_dir / "run_manifest.json"

    # Attempt to write the accumulated manifest data to disk
    try:
        # Open the manifest file in write mode with UTF-8 encoding
        with open(manifest_path, "w", encoding="utf-8") as f:
            # Serialize the dictionary to a formatted JSON string and write it
            json.dump(manifest_data, f, indent=4)
    # Catch any filesystem errors (e.g., permission denied, disk full)
    except OSError as exc:
        # Raise a fatal OSError if the audit trail cannot be preserved
        raise OSError(f"Failed to write run manifest to {manifest_path}: {exc}") from exc

    # Define the base name for the archive, matching the name of the output directory
    archive_base_name: str = str(run_output_dir.resolve())

    # Attempt to compress the entire output directory into a ZIP archive
    try:
        # Use shutil.make_archive to create a zip file containing the directory contents
        archive_path_str: str = shutil.make_archive(
            base_name=archive_base_name,
            format="zip",
            root_dir=str(run_output_dir.parent),
            base_dir=run_output_dir.name
        )
    # Catch any archiving errors
    except Exception as exc:
        # Raise a fatal OSError if the immutable deliverable cannot be created
        raise OSError(f"Failed to create compressed archive for {run_output_dir}: {exc}") from exc

    # Convert the returned string path into a pathlib.Path object
    archive_path: Path = Path(archive_path_str)

    # Log the successful creation of the immutable archive
    logger.info(f"Successfully created immutable artifact archive at: {archive_path}")

    # Return the path to the final compressed archive
    return archive_path

# -------------------------------------------------------------------------------------------------------------------------------
# Task 34, Helper: Construct mu_dict from persisted CMA artifacts
# -------------------------------------------------------------------------------------------------------------------------------

def _construct_mu_dict_from_artifacts(
    config: Dict[str, Any],
    run_output_dir: Path
) -> Dict[str, np.ndarray]:
    """
    Constructs the expected return mapping (mu_dict) by traversing the persisted
    cma.json artifacts for all 18 asset classes.

    This function implements the critical remediation blueprint, ensuring that
    Portfolio Construction (PC) agents receive mathematically valid Capital Market
    Assumptions (CMAs) rather than an empty placeholder dictionary.

    Parameters
    ----------
    config : Dict[str, Any]
        The validated study configuration dictionary containing the IPS asset classes
        and the PC method registry.
    run_output_dir : Path
        The root directory for the current run's artifacts.

    Returns
    -------
    Dict[str, np.ndarray]
        A dictionary mapping PC method slugs (where uses_cmas == True) to the
        18-element expected return vector.

    Raises
    ------
    KeyError
        If required configuration keys are missing.
    FileNotFoundError
        If any of the 18 required cma.json artifacts are missing from disk.
    ValueError
        If a cma.json artifact is unparseable or missing the 'final_estimate' key.
    """
    logger.info("Constructing mu_dict from persisted CMA artifacts.")

    # Step 1: Canonical Order Extraction
    try:
        asset_classes_dict: Dict[str, List[str]] = config["IPS_GOVERNANCE"]["ASSET_CLASSES"]
        pc_registry: Dict[str, Any] = config["REGISTRIES"]["PC_AGENTS"]["method_configs"]
    except KeyError as exc:
        raise KeyError(f"Configuration is missing required keys for mu_dict construction: {exc}") from exc

    # Flatten the dictionary into a canonical ordered list of 18 assets
    canonical_order: List[str] = []
    for category in ["equity", "fixed_income", "real_assets", "cash"]:
        canonical_order.extend(asset_classes_dict.get(category, []))

    # Extract the list of equity classes to route the directory paths correctly
    equity_classes: List[str] = asset_classes_dict.get("equity", [])

    # Step 2: Vector Initialization
    # Initialize a zeroed 18x1 NumPy array to hold the expected returns
    mu_vector: np.ndarray = np.zeros(18, dtype=np.float64)

    # Step 3: Artifact Traversal & Extraction
    for i, asset_class in enumerate(canonical_order):
        # Derive the filesystem slug for the asset class
        slug_lower: str = asset_class.lower().replace(" ", "_")
        slug: str = re.sub(r"[^a-z0-9_]", "", slug_lower)

        # Construct the path to the final CMA artifact
        # Equities go through the CMA Judge; others output directly from AC Agents
        if asset_class in equity_classes:
            cma_path: Path = run_output_dir / "cma_judge" / slug / "cma.json"
        else:
            cma_path: Path = run_output_dir / "ac_agents" / slug / "cma.json"

        # Fail-closed: Ensure the artifact exists
        if not cma_path.exists():
            raise FileNotFoundError(
                f"Fatal Error: Required CMA artifact missing for '{asset_class}' at {cma_path}. "
                f"Cannot construct expected return vector."
            )

        # Open and parse the JSON file
        try:
            with open(cma_path, "r", encoding="utf-8") as f:
                cma_data: Dict[str, Any] = json.load(f)

            # Extract the final_estimate float value
            final_estimate: float = float(cma_data["final_estimate"])

            # Assign this value to the corresponding index in the mu_vector
            mu_vector[i] = final_estimate

        except (json.JSONDecodeError, KeyError, ValueError) as exc:
            raise ValueError(f"Failed to parse or extract 'final_estimate' from {cma_path}: {exc}") from exc

    logger.debug(f"Successfully constructed 18-element mu_vector. Mean expected return: {np.mean(mu_vector):.4f}")

    # Step 4: Dictionary Population
    mu_dict: Dict[str, np.ndarray] = {}

    # Iterate through the PC registry
    for method_slug, method_config in pc_registry.items():
        # For every method where uses_cmas == True, assign the constructed mu_vector
        if method_config.get("uses_cmas", False):
            # Use a copy to prevent unintended shared-state mutations during optimization
            mu_dict[method_slug] = mu_vector.copy()

    logger.info(f"Successfully populated mu_dict for {len(mu_dict)} CMA-dependent PC methods.")

    return mu_dict

# -------------------------------------------------------------------------------------------------------------------------------
# Task 34, Master Orchestrator: End-to-End Pipeline DAG
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_34_end_to_end_pipeline(
    raw_config: Dict[str, Any],
    raw_dataframes: Dict[str, pd.DataFrame],
    client: Any,
    base_output_path: str = "./run_outputs"
) -> Path:
    """
    The Master Orchestrator for the Self-Driving Portfolio architecture.
    Executes a strictly sequential, fail-closed Directed Acyclic Graph (DAG)
    spanning Tasks 1 through 33, feature engineering, and skills generation.

    CRITICAL REMEDIATION: Integrates the `_construct_mu_dict_from_artifacts`
    helper prior to Phase 8, ensuring that the Portfolio Construction agents
    receive a mathematically valid expected return vector, eliminating the
    prohibited placeholder code.

    Parameters
    ----------
    raw_config : Dict[str, Any]
        The raw, unvalidated STUDY_CONFIG dictionary.
    raw_dataframes : Dict[str, pd.DataFrame]
        A dictionary mapping dataframe string identifiers to their raw pandas DataFrames.
    client : Any
        The authenticated OpenAI client instance.
    base_output_path : str, optional
        The root path where the unique run directory will be created. Defaults to "./run_outputs".

    Returns
    -------
    Path
        The absolute path to the final compressed, immutable archive file (.zip).

    Raises
    ------
    SystemExit
        If any unhandled exception occurs during the pipeline execution, enforcing
        a strict fail-closed halt (sys.exit(1)).
    """
    # Extract the current UTC timestamp to mark the exact start of the pipeline execution
    pipeline_start_utc: str = datetime.now(tz=timezone.utc).isoformat()

    # Generate a cryptographically random UUID to ensure the run directory is globally unique
    run_uuid: str = str(uuid.uuid4())

    # Format the timestamp to be filesystem-safe (removing colons and special characters)
    safe_timestamp: str = datetime.now(tz=timezone.utc).strftime("%Y%m%d_%H%M%S")

    # Construct the unique directory name combining the timestamp and the UUID
    run_dir_name: str = f"run_{safe_timestamp}_{run_uuid[:8]}"

    # Resolve the absolute path for the unique run output directory
    run_output_dir: Path = Path(base_output_path).resolve() / run_dir_name

    # Create the run output directory, including any necessary parent directories
    run_output_dir.mkdir(parents=True, exist_ok=True)

    # Initialize the master manifest dictionary to accumulate provenance metadata
    manifest_data: Dict[str, Any] = {
        "run_id": run_dir_name,
        "pipeline_start_utc": pipeline_start_utc,
        "status": "INITIALIZING",
        "data_provenance_hashes": {}
    }

    # Log the commencement of the master orchestrator
    logger.info(f"=== STARTING SELF-DRIVING PORTFOLIO PIPELINE (Run ID: {run_dir_name}) ===")

    # Wrap the entire DAG in a massive try-except block to guarantee fail-closed semantics
    try:
        # -------------------------------------------------------------------------------------------
        # PRE-FLIGHT: Compute cryptographic hashes of all raw ingested dataframes for provenance
        # -------------------------------------------------------------------------------------------
        logger.info("Computing SHA-256 provenance hashes for all raw ingested dataframes.")
        # Iterate through the dictionary of raw dataframes
        for df_name, df in raw_dataframes.items():
            # Serialize the dataframe to a CSV byte stream in memory
            csv_bytes: bytes = df.to_csv(index=True).encode('utf-8')
            # Compute the SHA-256 hash of the byte stream
            df_hash: str = hashlib.sha256(csv_bytes).hexdigest()
            # Store the hash in the manifest dictionary
            manifest_data["data_provenance_hashes"][df_name] = df_hash

        # ===========================================================================================
        # PHASE 1: Initialization and Configuration Freezing (Tasks 1–5)
        # ===========================================================================================
        logger.info("--- PHASE 1: Initialization and Configuration Freezing ---")

        # Step 1.1 (Task 1): Validate structure, enforce 18-asset invariant, and freeze config
        # Note: Assuming orchestrate_task_1_config_validation is available in the environment
        immutable_config, config_hash = orchestrate_task_1_config_validation(raw_config)
        # Store the cryptographic hash of the configuration in the manifest
        manifest_data["config_sha256"] = config_hash

        # Step 1.2 (Task 2): Strip legacy stochastic parameters and enforce GPT-5.2 API contract
        # Note: Assuming orchestrate_task_2_validate_gpt52_compatibility is available
        patched_config_1 = orchestrate_task_2_validate_gpt52_compatibility(dict(immutable_config))

        # Step 1.3 (Task 3): Mathematically lock the simple periodic return formula and annualization
        # Note: Assuming orchestrate_task_3_freeze_conventions is available
        patched_config_2 = orchestrate_task_3_freeze_conventions(patched_config_1)

        # Step 1.4 (Task 4): Translate qualitative IPS mandates into strict mathematical boundaries
        # Note: Assuming orchestrate_task_4_validate_ips_computability is available
        patched_config_3 = orchestrate_task_4_validate_ips_computability(patched_config_2)

        # Step 1.5 (Task 5): Verify 60/40 benchmark inputs and project weights into 18-D space
        # Note: Assuming orchestrate_task_5_validate_benchmark is available
        r_benchmark, w_b_18_dim = orchestrate_task_5_validate_benchmark(
            df_benchmark_60_40_raw=raw_dataframes["df_benchmark_60_40_raw"],
            config=patched_config_3
        )

        # ===========================================================================================
        # PHASE 2: Prompt Inventory and LLM Boundary Enforcement (Tasks 6–8)
        # ===========================================================================================
        logger.info("--- PHASE 2: Prompt Inventory and LLM Boundary Enforcement ---")

        # Step 2.1 (Task 6): Verify existence of all 15 templates and verbatim provenance of Exhibits 3/4
        # Note: Assuming orchestrate_task_6_validate_prompt_inventory is available
        orchestrate_task_6_validate_prompt_inventory(patched_config_3)

        # Step 2.2 (Task 7): Scan prompts for arithmetic prohibitions and cross-reference tool placeholders
        # Note: Assuming orchestrate_task_7_validate_prompt_tool_boundaries is available
        orchestrate_task_7_validate_prompt_tool_boundaries(patched_config_3)

        # Step 2.3 (Task 8): Enforce 'developer' role and validate the token budget schedule
        # Note: Assuming orchestrate_task_8_validate_gpt52_conventions is available
        final_config = orchestrate_task_8_validate_gpt52_conventions(patched_config_3)

        # Record the final, fully patched configuration and random seeds in the manifest
        manifest_data["final_configuration"] = final_config
        manifest_data["random_seeds"] = final_config["RANDOM_SEEDS"]

        # ===========================================================================================
        # PHASE 3: Data Ingestion and Foundational Cleansing (Tasks 9–15)
        # ===========================================================================================
        logger.info("--- PHASE 3: Data Ingestion and Foundational Cleansing ---")

        # Step 3.1 (Task 9): Enforce schema compliance, index uniqueness, and point-in-time truncation
        # Note: Assuming orchestrate_task_9_validate_ingested_dataframes is available
        validated_dfs = orchestrate_task_9_validate_ingested_dataframes(raw_dataframes, final_config)

        # Step 3.2 (Task 10): Strip timezones, snap to midnight, and align daily factors to monthly
        # Note: Assuming orchestrate_task_10_cleanse_time_indices is available
        cleansed_dfs = orchestrate_task_10_cleanse_time_indices(validated_dfs, final_config, run_output_dir)

        # Step 3.3 (Task 11): Validate positive total returns and construct the canonical T x 18 matrix
        # Note: Assuming orchestrate_task_11_construct_monthly_returns_panel is available
        returns_matrix = orchestrate_task_11_construct_monthly_returns_panel(
            df_total_return_raw=cleansed_dfs["df_total_return_raw"],
            universe_map=final_config["REGISTRIES"]["universe_map"],
            config=final_config
        )

        # Step 3.4 (Task 12): Enforce 1-month GDP publication lag and apply bounded forward-fill
        # Note: Assuming orchestrate_task_12_cleanse_macro_series is available
        df_macro_clean = orchestrate_task_12_cleanse_macro_series(
            df_macro_raw=cleansed_dfs["df_macro_raw"],
            config=final_config,
            run_output_dir=run_output_dir
        )

        # Step 3.5 (Task 13): Nullify equity fields for non-equities and synchronize feature space
        # Note: Assuming orchestrate_task_13_cleanse_fundamentals_and_signals is available
        df_fundamentals_clean, df_signals_clean = orchestrate_task_13_cleanse_fundamentals_and_signals(
            df_fundamentals_raw=cleansed_dfs["df_fundamentals_raw"],
            df_signals_raw=cleansed_dfs["df_signals_raw"],
            df_macro_clean=df_macro_clean,
            config=final_config
        )

        # Step 3.6 (Task 14): Finalize benchmark return series and align perfectly with returns_matrix
        # Note: Assuming orchestrate_task_14_cleanse_and_compute_benchmark is available
        r_benchmark_aligned = orchestrate_task_14_cleanse_and_compute_benchmark(
            df_benchmark_60_40_raw=cleansed_dfs["df_benchmark_60_40_raw"],
            returns_matrix=returns_matrix,
            config=final_config
        )

        # Step 3.7 (Task 15): Verify all 18 mapped tickers exist with sufficient temporal depth
        # Note: Assuming orchestrate_task_15_validate_universe_map is available
        integrity_report = orchestrate_task_15_validate_universe_map(
            universe_map=final_config["REGISTRIES"]["universe_map"],
            df_total_return_raw=cleansed_dfs["df_total_return_raw"],
            df_ohlcv_raw=cleansed_dfs["df_ohlcv_raw"],
            config=final_config,
            run_output_dir=run_output_dir
        )

        # ===========================================================================================
        # PHASE 4: Deterministic Feature Engineering
        # ===========================================================================================
        logger.info("--- PHASE 4: Deterministic Feature Engineering ---")

        # Step 4.2: Invoke the feature creation module to synthesize missing columns (e.g., CPI MoM)
        # Note: Assuming orchestrate_feature_creation is available
        engineered_dfs = orchestrate_feature_creation(
            df_macro_raw=df_macro_clean,
            df_cpi_index_raw=cleansed_dfs["df_cpi_index_raw"],
            df_total_return_raw=cleansed_dfs["df_total_return_raw"],
            df_benchmark_factors_raw=cleansed_dfs["df_benchmark_factors_raw"],
            is_rf_annualized=True # Assuming the raw RF is annualized based on standard conventions
        )

        # Step 4.3: Overwrite the pipeline's state variables with the fully engineered dataframes
        df_macro_final = engineered_dfs["df_macro_processed"]
        df_total_return_final = engineered_dfs["df_total_return_processed"]
        df_factors_monthly_final = engineered_dfs["df_benchmark_factors_monthly"]

        # ===========================================================================================
        # PHASE 5: Output Contracts and Skills Registry Generation
        # ===========================================================================================
        logger.info("--- PHASE 5: Output Contracts and Skills Registry Generation ---")

        # Step 5.1 (Task 16): Generate strict JSON Schemas and return the bound validation callable
        # Note: Assuming orchestrate_task_16_define_output_contracts is available
        validate_artifact_fn = orchestrate_task_16_define_output_contracts(final_config)

        # Step 5.2: Generate definitive markdown contracts defining the 6-9 skills per agent
        # Note: Assuming generate_and_persist_agent_skills_registry is available
        generate_and_persist_agent_skills_registry(base_dir=run_output_dir)

        # ===========================================================================================
        # PHASE 6: Macro Regime Synthesis (Tasks 17 & 20)
        # ===========================================================================================
        logger.info("--- PHASE 6: Macro Regime Synthesis ---")

        # Step 6.1 (Task 20): Deterministically compute z-scores, map to regime, and compute confidence
        # Note: Assuming orchestrate_task_20_macro_scoring_and_classification is available
        macro_view_dict = orchestrate_task_20_macro_scoring_and_classification(
            df_macro_raw=df_macro_final,
            config=final_config
        )

        # Step 6.2 (Task 17): Execute ReAct loop for LLM to synthesize the fiduciary narrative
        # Note: Assuming orchestrate_task_17_macro_agent is available
        macro_json_path, macro_md_path = orchestrate_task_17_macro_agent(
            df_macro_raw=df_macro_final,
            config=final_config,
            run_output_dir=run_output_dir,
            client=client,
            validate_artifact_fn=validate_artifact_fn
        )

        # ===========================================================================================
        # PHASE 7: Capital Market Assumptions Generation (Tasks 18, 21, 22, 23, 24)
        # ===========================================================================================
        logger.info("--- PHASE 7: Capital Market Assumptions Generation ---")

        # Step 7.1 (Task 24): Compute the 18x18 Ledoit-Wolf shrinkage covariance matrix (Sigma)
        # Note: Assuming orchestrate_task_24_covariance_agent is available
        sigma = orchestrate_task_24_covariance_agent(
            df_total_return_raw=df_total_return_final,
            config=final_config,
            run_output_dir=run_output_dir
        )

        # Step 7.3 (Task 18): Dispatch 18 parallel ReAct loops to generate cma_methods.json
        # Note: Assuming orchestrate_task_18_ac_agents is available
        ac_artifacts = orchestrate_task_18_ac_agents(
            df_total_return_raw=df_total_return_final,
            df_signals_raw=df_signals_clean,
            df_fundamentals_raw=df_fundamentals_clean,
            df_survey_cma_raw=cleansed_dfs["df_survey_cma_raw"],
            df_fixed_income_curves_spreads_raw=cleansed_dfs["df_fixed_income_curves_spreads_raw"],
            df_benchmark_factors_raw=cleansed_dfs["df_benchmark_factors_raw"],
            macro_view_path=macro_json_path,
            sigma=sigma,
            config=final_config,
            run_output_dir=run_output_dir,
            client=client,
            validate_artifact_fn=validate_artifact_fn
        )

        # Step 7.4 (Task 22): Execute CMA Judge for each equity asset class to select final estimates
        # Note: Assuming orchestrate_task_22_cma_judge is available
        equity_classes: List[str] = final_config["IPS_GOVERNANCE"]["ASSET_CLASSES"]["equity"]
        cma_judge_artifacts: Dict[str, str] = {}
        for asset_class in equity_classes:
            cma_path = orchestrate_task_22_cma_judge(
                asset_class=asset_class,
                config=final_config,
                run_output_dir=run_output_dir,
                client=client,
                validate_artifact_fn=validate_artifact_fn
            )
            cma_judge_artifacts[asset_class] = cma_path

        # ===========================================================================================
        # PHASE 8: Multi-Agent Strategy Review Pipeline (Tasks 19, 25–31)
        # ===========================================================================================
        logger.info("--- PHASE 8: Multi-Agent Strategy Review Pipeline ---")

        # Extract the current risk-free rate for optimization and reporting
        rf_current: float = df_factors_monthly_final["rf"].iloc[-1]

        # Extract the temporal index from the returns matrix for factor regression alignment
        # We reconstruct the DatetimeIndex from the cleansed total return dataframe
        returns_index: pd.DatetimeIndex = df_total_return_final.index.get_level_values("date").unique().sort_values()
        # Slice the index to match the trailing window length of the returns matrix
        returns_index = returns_index[-returns_matrix.shape[0]:]

        # CRITICAL REMEDIATION: Construct the mu_dict dynamically from the persisted CMA artifacts
        # This replaces the prohibited placeholder code and ensures mathematical validity
        mu_dict: Dict[str, np.ndarray] = _construct_mu_dict_from_artifacts(
            config=final_config,
            run_output_dir=run_output_dir
        )

        # Execute the massive AutoGen FSM topology (Phases A through G)
        # Note: Assuming orchestrate_task_19_strategy_review is available
        final_weights_path, board_memo_path = orchestrate_task_19_strategy_review(
            config=final_config,
            sigma=sigma,
            mu_dict=mu_dict, # Now passing the fully populated, mathematically accurate dictionary
            rf=rf_current,
            benchmark_weights=w_b_18_dim,
            returns_matrix=returns_matrix,
            returns_index=returns_index,
            factors_df=df_factors_monthly_final,
            macro_view_path=Path(macro_json_path),
            cma_dir=run_output_dir / "cma_judge",
            run_output_dir=run_output_dir,
            openai_api_key=final_config["INFRASTRUCTURE"]["PRIMARY_LLM"]["API_KEY"],
            validate_artifact_fn=validate_artifact_fn
        )

        # ===========================================================================================
        # PHASE 9: Fiduciary Reporting and Diagnostics (Tasks 32–33)
        # ===========================================================================================
        logger.info("--- PHASE 9: Fiduciary Reporting and Diagnostics ---")

        # Step 9.1 (Task 32): Reproduce core tables (Exhibits 8-11)
        # Note: Assuming orchestrate_task_32_reproduce_core_tables is available
        exhibits_dict = orchestrate_task_32_reproduce_core_tables(
            config=final_config,
            run_output_dir=run_output_dir,
            sigma=sigma,
            returns_matrix=returns_matrix,
            rf=rf_current,
            benchmark_weights=w_b_18_dim,
            universe_map=final_config["REGISTRIES"]["universe_map"]
        )

        # Step 9.2 (Task 33): Compute realized returns, metrics, and verify ex-post IPS compliance
        # Note: Assuming orchestrate_task_33_backtest_diagnostics is available
        diagnostics_report = orchestrate_task_33_backtest_diagnostics(
            config=final_config,
            run_output_dir=run_output_dir,
            returns_matrix=returns_matrix,
            returns_index=returns_index,
            benchmark_returns=r_benchmark_aligned,
            rf=rf_current
        )

        # Update manifest status to SUCCESS
        manifest_data["status"] = "SUCCESS"

    # Catch ANY exception thrown during the pipeline execution to enforce fail-closed semantics
    except Exception as exc:
        # Extract the current UTC timestamp to mark the exact time of failure
        pipeline_fail_utc: str = datetime.now(tz=timezone.utc).isoformat()

        # Update the manifest dictionary with the failure state and error message
        manifest_data["status"] = "FAILED"
        manifest_data["failure_utc"] = pipeline_fail_utc
        manifest_data["failure_message"] = str(exc)

        # Log the critical failure to the audit trail
        logger.critical(f"PIPELINE FAILED: {exc}. Writing failure manifest and halting.")

        # Attempt to write the failure manifest to disk before exiting
        try:
            finalize_provenance_and_manifest(run_output_dir, manifest_data)
        except Exception as manifest_exc:
            logger.critical(f"Failed to write failure manifest: {manifest_exc}")

        # Invoke sys.exit(1) to immediately halt the Python process, guaranteeing a fail-closed state
        sys.exit(1)

    # ===========================================================================================
    # PHASE 10: Provenance and Manifest Finalization
    # ===========================================================================================
    logger.info("--- PHASE 10: Provenance and Manifest Finalization ---")

    # Extract the current UTC timestamp to mark the exact end of the successful pipeline execution
    pipeline_end_utc: str = datetime.now(tz=timezone.utc).isoformat()
    manifest_data["pipeline_end_utc"] = pipeline_end_utc

    # Invoke the helper function to serialize the manifest and compress the artifact directory
    archive_path: Path = finalize_provenance_and_manifest(run_output_dir, manifest_data)

    # Log the absolute completion of the master orchestrator
    logger.info(f"=== SELF-DRIVING PORTFOLIO PIPELINE COMPLETED SUCCESSFULLY ===")
    logger.info(f"Final Archive: {archive_path}")

    # Return the path to the immutable, compressed deliverable
    return archive_path
````

````python
# Task 35 — Robustness analysis: data and modeling sensitivity (must use orchestrator)

# ==============================================================================
# Task 35: Robustness analysis: data and modeling sensitivity
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 35, Step 1: Generate alternative frozen parameter sets (Configurations)
# -------------------------------------------------------------------------------------------------------------------------------

def generate_robustness_configurations(
    base_config: Dict[str, Any]
) -> List[Tuple[str, Dict[str, Any]]]:
    """
    Generates a suite of mutated configuration dictionaries for sensitivity analysis.
    Applies one-at-a-time (ceteris paribus) perturbations to covariance estimation
    parameters and macroeconomic scoring weights.

    Parameters
    ----------
    base_config : Dict[str, Any]
        The validated, baseline study configuration dictionary.

    Returns
    -------
    List[Tuple[str, Dict[str, Any]]]
        A list of tuples, where each tuple contains a deterministic run ID string
        and the deeply copied, mutated configuration dictionary.

    Raises
    ------
    KeyError
        If the required baseline configuration keys are missing.
    """
    logger.info("Generating robustness configurations for sensitivity analysis.")

    configurations: List[Tuple[str, Dict[str, Any]]] = []

    try:
        # Extract baseline parameters to ensure they exist
        cov_params: Dict[str, Any] = base_config["METHODOLOGY_PARAMS"]["COVARIANCE_ESTIMATION"]
        baseline_window: int = cov_params["window_months"]
        baseline_estimator: str = cov_params["estimator"]

        macro_weights: Dict[str, Dict[str, float]] = base_config["METHODOLOGY_PARAMS"]["MACRO_SCORING_WEIGHTS"]
        baseline_gdp_weight: float = macro_weights["growth"]["real_gdp_growth"]
        baseline_nfp_weight: float = macro_weights["growth"]["nonfarm_payrolls_mom"]
    except KeyError as exc:
        raise KeyError(f"Base configuration is missing required keys for perturbation: {exc}") from exc

    # 1. Perturb Covariance Window (Short: 60 months)
    run_id_window_short: str = "robustness_cov_window_60"
    config_window_short: Dict[str, Any] = copy.deepcopy(base_config)
    config_window_short["METHODOLOGY_PARAMS"]["COVARIANCE_ESTIMATION"]["window_months"] = 60
    configurations.append((run_id_window_short, config_window_short))

    # 2. Perturb Covariance Window (Long: 180 months)
    run_id_window_long: str = "robustness_cov_window_180"
    config_window_long: Dict[str, Any] = copy.deepcopy(base_config)
    config_window_long["METHODOLOGY_PARAMS"]["COVARIANCE_ESTIMATION"]["window_months"] = 180
    configurations.append((run_id_window_long, config_window_long))

    # 3. Perturb Covariance Estimator (Sample Covariance)
    run_id_est_sample: str = "robustness_cov_estimator_sample"
    config_est_sample: Dict[str, Any] = copy.deepcopy(base_config)
    config_est_sample["METHODOLOGY_PARAMS"]["COVARIANCE_ESTIMATION"]["estimator"] = "sample"
    configurations.append((run_id_est_sample, config_est_sample))

    # 4. Perturb Macro Scoring Weights (GDP Heavy)
    # Shift GDP weight up by 0.2, reduce NFP by 0.2 to maintain sum-to-one relative weighting
    run_id_macro_gdp: str = "robustness_macro_growth_gdp_heavy"
    config_macro_gdp: Dict[str, Any] = copy.deepcopy(base_config)
    new_gdp_weight: float = min(1.0, baseline_gdp_weight + 0.2)
    new_nfp_weight: float = max(0.0, baseline_nfp_weight - 0.2)
    config_macro_gdp["METHODOLOGY_PARAMS"]["MACRO_SCORING_WEIGHTS"]["growth"]["real_gdp_growth"] = new_gdp_weight
    config_macro_gdp["METHODOLOGY_PARAMS"]["MACRO_SCORING_WEIGHTS"]["growth"]["nonfarm_payrolls_mom"] = new_nfp_weight
    configurations.append((run_id_macro_gdp, config_macro_gdp))

    # 5. Perturb Macro Scoring Weights (NFP Heavy)
    # Shift NFP weight up by 0.2, reduce GDP by 0.2
    run_id_macro_nfp: str = "robustness_macro_growth_nfp_heavy"
    config_macro_nfp: Dict[str, Any] = copy.deepcopy(base_config)
    new_nfp_weight_alt: float = min(1.0, baseline_nfp_weight + 0.2)
    new_gdp_weight_alt: float = max(0.0, baseline_gdp_weight - 0.2)
    config_macro_nfp["METHODOLOGY_PARAMS"]["MACRO_SCORING_WEIGHTS"]["growth"]["real_gdp_growth"] = new_gdp_weight_alt
    config_macro_nfp["METHODOLOGY_PARAMS"]["MACRO_SCORING_WEIGHTS"]["growth"]["nonfarm_payrolls_mom"] = new_nfp_weight_alt
    configurations.append((run_id_macro_nfp, config_macro_nfp))

    logger.info(f"Successfully generated {len(configurations)} robustness configurations.")
    return configurations

# -------------------------------------------------------------------------------------------------------------------------------
# Task 35, Step 2: Execute isolated robustness trials
# -------------------------------------------------------------------------------------------------------------------------------

def execute_robustness_trials(
    configurations: List[Tuple[str, Dict[str, Any]]],
    raw_dataframes: Dict[str, pd.DataFrame],
    client: Any,
    base_output_path: Path
) -> Dict[str, Path]:
    """
    Executes the master orchestrator sequentially for each perturbed configuration.
    Isolates the output of each trial into a specific subdirectory to prevent
    artifact collision and state leakage.

    Parameters
    ----------
    configurations : List[Tuple[str, Dict[str, Any]]]
        The list of run IDs and their corresponding mutated configurations.
    raw_dataframes : Dict[str, pd.DataFrame]
        The raw, unmutated dataframes required for pipeline ingestion.
    client : Any
        The authenticated OpenAI client.
    base_output_path : Path
        The root directory where the robustness trial folders will be created.

    Returns
    -------
    Dict[str, Path]
        A dictionary mapping the run ID to the absolute path of its generated
        artifact directory.
    """
    logger.info(f"Executing {len(configurations)} isolated robustness trials.")

    trial_directories: Dict[str, Path] = {}

    for run_id, config in configurations:
        logger.info(f"--- Starting Robustness Trial: {run_id} ---")

        # Define a strict, isolated output path for this specific trial
        trial_output_path: Path = base_output_path / run_id
        trial_output_path.mkdir(parents=True, exist_ok=True)

        try:
            # Invoke the master orchestrator (Task 34)
            # We pass the isolated trial_output_path as the base directory
            archive_path: Path = orchestrate_task_34_end_to_end_pipeline(
                raw_config=config,
                raw_dataframes=raw_dataframes,
                client=client,
                base_output_path=str(trial_output_path)
            )

            # The orchestrator creates a UUID subfolder. We need to locate it.
            # We assume the archive name matches the UUID folder name.
            run_dir_name: str = archive_path.stem
            actual_run_dir: Path = trial_output_path / run_dir_name

            if actual_run_dir.exists():
                trial_directories[run_id] = actual_run_dir
                logger.info(f"Trial '{run_id}' completed successfully. Artifacts at: {actual_run_dir}")
            else:
                logger.error(f"Trial '{run_id}' completed, but artifact directory {actual_run_dir} not found.")

        except Exception as exc:
            # Graceful degradation: Log the failure but continue to the next trial
            logger.error(f"Robustness Trial '{run_id}' failed with exception: {exc}")
            # We do not add failed runs to the trial_directories dict

    logger.info(f"Completed robustness execution. {len(trial_directories)}/{len(configurations)} trials succeeded.")
    return trial_directories

# -------------------------------------------------------------------------------------------------------------------------------
# Task 35, Step 3: Compute mathematical drift metrics
# -------------------------------------------------------------------------------------------------------------------------------

def compute_robustness_metrics(
    baseline_dir: Path,
    trial_directories: Dict[str, Path],
    asset_classes: List[str]
) -> Dict[str, Dict[str, Any]]:
    """
    Computes strict mathematical drift metrics by comparing the artifacts of
    each perturbed trial against the baseline run.

    CRITICAL REMEDIATION: Eradicates blind array subtraction. Extracts
    'asset_class_labels' from the JSON artifacts and enforces strict
    cross-sectional alignment via dictionary mapping before computing the
    L1 distance. This guarantees mathematical fidelity even if a trial
    inadvertently reorders the asset vector.

    Metrics computed:
    1. L1 Distance of final weights: sum(|w_trial - w_base|)
    2. Regime Flip: Boolean indicating if the macro regime changed.
    3. Top-5 Overlap: Fraction of shared methods in the PC shortlist.
    4. CMA Mean Drift: Mean absolute difference in final CMA estimates.
    5. IPS Compliance: Boolean indicating if the final ensemble remained compliant.

    Parameters
    ----------
    baseline_dir : Path
        The artifact directory of the unperturbed baseline run.
    trial_directories : Dict[str, Path]
        Dictionary mapping run IDs to their artifact directories.
    asset_classes : List[str]
        The canonical list of 18 asset classes for CMA traversal and weight alignment.

    Returns
    -------
    Dict[str, Dict[str, Any]]
        A nested dictionary mapping run IDs to their computed drift metrics.
    """
    logger.info("Computing mathematical drift metrics for robustness trials with strict label alignment.")

    metrics_report: Dict[str, Dict[str, Any]] = {}

    # --- Helper to safely load JSON ---
    def _load_json(path: Path) -> Dict[str, Any]:
        if not path.exists():
            return {}
        try:
            with open(path, "r", encoding="utf-8") as f:
                return json.load(f)
        except json.JSONDecodeError:
            return {}

    # --- Load Baseline Artifacts ---
    # 1. Baseline Macro Regime
    base_macro: Dict[str, Any] = _load_json(baseline_dir / "macro" / "macro-view.json")
    base_regime: str = base_macro.get("regime", "UNKNOWN")

    # 2. Baseline Top-5 Selection
    base_top5_data: Dict[str, Any] = _load_json(baseline_dir / "top5_revision" / "top5_selection.json")
    base_top5: Set[str] = set(base_top5_data.get("top_5_methods", []))

    # 3. Baseline Final Weights and Labels
    base_weights_data: Dict[str, Any] = _load_json(baseline_dir / "cio" / "final_weights.json")
    base_weights_list: List[float] = base_weights_data.get("weights", [])
    base_labels_list: List[str] = base_weights_data.get("asset_class_labels", [])

    # Construct the baseline weight dictionary for safe alignment
    base_weight_dict: Dict[str, float] = {}
    if len(base_weights_list) == 18 and len(base_labels_list) == 18:
        base_weight_dict = dict(zip(base_labels_list, base_weights_list))
    else:
        logger.warning("Baseline final_weights.json is malformed or missing labels. L1 distance will be NaN.")

    # 4. Baseline CMA Estimates
    base_cmas: Dict[str, float] = {}
    for ac in asset_classes:
        slug: str = __import__("re").sub(r"[^a-z0-9_]", "", ac.lower().replace(" ", "_"))
        # Check both possible locations (CMA Judge for equities, AC Agents for others)
        cma_path_judge: Path = baseline_dir / "cma_judge" / slug / "cma.json"
        cma_path_ac: Path = baseline_dir / "ac_agents" / slug / "cma.json"
        cma_data: Dict[str, Any] = _load_json(cma_path_judge) if cma_path_judge.exists() else _load_json(cma_path_ac)
        base_cmas[ac] = float(cma_data.get("final_estimate", np.nan))

    # --- Compute Metrics per Trial ---
    for run_id, trial_dir in trial_directories.items():
        trial_metrics: Dict[str, Any] = {}

        # 1. Regime Flip
        trial_macro: Dict[str, Any] = _load_json(trial_dir / "macro" / "macro-view.json")
        trial_regime: str = trial_macro.get("regime", "UNKNOWN_TRIAL")
        trial_metrics["regime_flip"] = bool(trial_regime != base_regime)
        trial_metrics["trial_regime"] = trial_regime

        # 2. Top-5 Overlap
        trial_top5_data: Dict[str, Any] = _load_json(trial_dir / "top5_revision" / "top5_selection.json")
        trial_top5: Set[str] = set(trial_top5_data.get("top_5_methods", []))
        if base_top5 and trial_top5:
            overlap_fraction: float = len(base_top5.intersection(trial_top5)) / 5.0
        else:
            overlap_fraction = np.nan
        trial_metrics["top5_overlap_fraction"] = float(overlap_fraction)

        # 3. L1 Distance of Final Weights (Strictly Aligned)
        trial_weights_data: Dict[str, Any] = _load_json(trial_dir / "cio" / "final_weights.json")
        trial_weights_list: List[float] = trial_weights_data.get("weights", [])
        trial_labels_list: List[str] = trial_weights_data.get("asset_class_labels", [])

        # Verify that both baseline and trial dictionaries were successfully populated
        if base_weight_dict and len(trial_weights_list) == 18 and len(trial_labels_list) == 18:
            # Verify strict set equality of the labels to prevent silent dropping of assets
            if set(base_labels_list) == set(trial_labels_list):
                # Construct the trial weight dictionary
                trial_weight_dict: Dict[str, float] = dict(zip(trial_labels_list, trial_weights_list))

                # Extract aligned arrays using the canonical asset_classes list
                aligned_base_arr: np.ndarray = np.array([base_weight_dict[ac] for ac in asset_classes], dtype=np.float64)
                aligned_trial_arr: np.ndarray = np.array([trial_weight_dict[ac] for ac in asset_classes], dtype=np.float64)

                # Equation: L1 = sum(|w_trial - w_base|)
                l1_dist: float = float(np.sum(np.abs(aligned_trial_arr - aligned_base_arr)))
                # Round to 6 decimal places to eliminate floating-point noise
                trial_metrics["final_weight_l1_distance"] = round(l1_dist, 6)
            else:
                logger.error(f"Label mismatch in trial {run_id}. Cannot compute L1 distance safely.")
                trial_metrics["final_weight_l1_distance"] = np.nan
        else:
            trial_metrics["final_weight_l1_distance"] = np.nan

        # 4. CMA Mean Drift
        cma_drifts: List[float] = []
        for ac in asset_classes:
            slug: str = __import__("re").sub(r"[^a-z0-9_]", "", ac.lower().replace(" ", "_"))
            cma_path_judge: Path = trial_dir / "cma_judge" / slug / "cma.json"
            cma_path_ac: Path = trial_dir / "ac_agents" / slug / "cma.json"
            cma_data: Dict[str, Any] = _load_json(cma_path_judge) if cma_path_judge.exists() else _load_json(cma_path_ac)

            trial_est: float = float(cma_data.get("final_estimate", np.nan))
            base_est: float = base_cmas.get(ac, np.nan)

            if not np.isnan(trial_est) and not np.isnan(base_est):
                cma_drifts.append(abs(trial_est - base_est))

        if cma_drifts:
            trial_metrics["cma_mean_absolute_drift"] = float(np.mean(cma_drifts))
        else:
            trial_metrics["cma_mean_absolute_drift"] = np.nan

        # 5. IPS Compliance
        # If final_weights.json exists and has range_constraint_verified=True, it passed the CIO gate
        trial_metrics["ips_compliant"] = bool(trial_weights_data.get("range_constraint_verified", False))

        metrics_report[run_id] = trial_metrics
        logger.debug(f"Computed metrics for {run_id}: L1={trial_metrics['final_weight_l1_distance']}, Flip={trial_metrics['regime_flip']}")

    return metrics_report


def generate_robustness_report(
    metrics_report: Dict[str, Dict[str, Any]],
    output_dir: Path
) -> Path:
    """
    Synthesizes the computed mathematical drift metrics into a structured,
    actionable fiduciary report. Ranks the parameter perturbations by their
    impact magnitude (L1 distance) and explicitly identifies assumptions that
    drive regime flips and allocation instability.

    Parameters
    ----------
    metrics_report : Dict[str, Dict[str, Any]]
        The nested dictionary mapping run IDs to their computed drift metrics,
        produced by compute_robustness_metrics.
    output_dir : Path
        The directory where the robustness_report.json will be saved.

    Returns
    -------
    Path
        The absolute path to the written JSON report artifact.

    Raises
    ------
    ValueError
        If the metrics_report is empty, indicating no trials succeeded.
    OSError
        If the report cannot be written to disk due to filesystem permissions.
    """
    logger.info("Generating structured robustness and sensitivity report.")

    # Guard against an empty metrics report (e.g., if all trials failed)
    if not metrics_report:
        raise ValueError("Cannot generate robustness report: metrics_report is empty.")

    # Initialize lists to hold the extracted data for ranking and filtering
    valid_l1_distances: List[Tuple[str, float]] = []
    regime_flips: List[str] = []
    non_compliant_runs: List[str] = []

    # Iterate through the metrics report to extract and categorize the data
    for run_id, metrics in metrics_report.items():
        # Extract L1 distance, handling potential NaNs from failed weight generations
        l1_dist: float = metrics.get("final_weight_l1_distance", np.nan)
        if not np.isnan(l1_dist):
            valid_l1_distances.append((run_id, float(l1_dist)))

        # Identify runs that caused a macro regime flip
        if metrics.get("regime_flip", False):
            regime_flips.append(run_id)

        # Identify runs that resulted in an IPS non-compliant final ensemble
        if not metrics.get("ips_compliant", True):
            non_compliant_runs.append(run_id)

    # Sort the valid runs descending by L1 distance to rank sensitivity
    # The run with the highest L1 distance represents the most fragile assumption
    valid_l1_distances.sort(key=lambda x: x[1], reverse=True)

    # Construct the ranked list of dictionaries for the JSON payload
    rankings: List[Dict[str, Any]] = [
        {"run_id": run_id, "l1_distance": dist, "rank": i + 1}
        for i, (run_id, dist) in enumerate(valid_l1_distances)
    ]

    # Identify the single most sensitive run (if any valid runs exist)
    most_sensitive_run: str = valid_l1_distances[0][0] if valid_l1_distances else "None"
    max_l1_distance: float = valid_l1_distances[0][1] if valid_l1_distances else 0.0

    # Construct the final structured JSON payload
    report_payload: Dict[str, Any] = {
        "summary": {
            "total_trials_evaluated": len(metrics_report),
            "trials_with_valid_weights": len(valid_l1_distances),
            "most_sensitive_assumption_run_id": most_sensitive_run,
            "maximum_l1_weight_drift": max_l1_distance,
            "total_regime_flips": len(regime_flips),
            "total_ips_violations": len(non_compliant_runs)
        },
        "sensitivity_rankings": rankings,
        "regime_instability_drivers": regime_flips,
        "ips_compliance_failures": non_compliant_runs,
        "raw_metrics": metrics_report
    }

    # Ensure the output directory exists
    output_dir.mkdir(parents=True, exist_ok=True)

    # Define the standardized filename
    report_path: Path = output_dir / "robustness_report.json"

    # Attempt to write the report to disk
    try:
        with open(report_path, "w", encoding="utf-8") as f:
            json.dump(report_payload, f, indent=4)
    except OSError as exc:
        # Raise a fatal OSError if the audit trail cannot be preserved
        raise OSError(f"Failed to write robustness report to {report_path}: {exc}") from exc

    logger.info(f"Successfully generated and persisted robustness report to {report_path}.")

    # Return the absolute path to the artifact
    return report_path.resolve()

# -------------------------------------------------------------------------------------------------------------------------------
# Task 35, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_35_data_modeling_robustness(
    base_config: Dict[str, Any],
    raw_dataframes: Dict[str, pd.DataFrame],
    client: Any,
    baseline_dir: Path,
    base_output_path: Path
) -> Path:
    """
    The Master Orchestrator for Task 35. Executes the full data and modeling
    sensitivity analysis pipeline.

    Generates perturbed configurations, executes isolated end-to-end trials,
    computes mathematical drift metrics against the baseline, and synthesizes
    the findings into a structured fiduciary report.

    Parameters
    ----------
    base_config : Dict[str, Any]
        The validated, baseline study configuration dictionary.
    raw_dataframes : Dict[str, pd.DataFrame]
        The raw, unmutated dataframes required for pipeline ingestion.
    client : Any
        The authenticated OpenAI client.
    baseline_dir : Path
        The artifact directory of the unperturbed baseline run. Must exist.
    base_output_path : Path
        The root directory where the robustness trial folders will be created.

    Returns
    -------
    Path
        The absolute path to the final robustness_report.json artifact.

    Raises
    ------
    FileNotFoundError
        If the baseline_dir does not exist, making comparison impossible.
    RuntimeError
        If any unhandled exception occurs during the orchestration flow,
        enforcing a strict fail-closed halt.
    """
    logger.info("=== STARTING TASK 35: DATA AND MODELING ROBUSTNESS ANALYSIS ===")

    # Pre-flight check: Ensure the baseline directory exists for comparison
    if not baseline_dir.exists() or not baseline_dir.is_dir():
        raise FileNotFoundError(
            f"Fatal Error: Baseline directory {baseline_dir} does not exist. "
            f"Robustness analysis requires a completed baseline run for comparison."
        )

    # Extract the canonical asset classes for CMA drift computation
    try:
        asset_classes_dict: Dict[str, List[str]] = base_config["IPS_GOVERNANCE"]["ASSET_CLASSES"]
        canonical_order: List[str] = []
        for category in ["equity", "fixed_income", "real_assets", "cash"]:
            canonical_order.extend(asset_classes_dict.get(category, []))
    except KeyError as exc:
        raise RuntimeError(f"Configuration is missing required keys for Task 35: {exc}") from exc

    # Wrap the DAG in a try/except block to guarantee fail-closed semantics
    try:
        # Step 1: Generate alternative frozen parameter sets
        # Note: Assuming generate_robustness_configurations is available in the environment
        configurations: List[Tuple[str, Dict[str, Any]]] = generate_robustness_configurations(
            base_config=base_config
        )

        # Step 2: Execute isolated robustness trials
        # Note: Assuming execute_robustness_trials is available in the environment
        trial_directories: Dict[str, Path] = execute_robustness_trials(
            configurations=configurations,
            raw_dataframes=raw_dataframes,
            client=client,
            base_output_path=base_output_path
        )

        # Guard: Ensure at least one trial succeeded before attempting metric computation
        if not trial_directories:
            raise RuntimeError("All robustness trials failed. Cannot compute drift metrics.")

        # Step 3a: Compute mathematical drift metrics
        # Note: Assuming compute_robustness_metrics is available in the environment
        metrics_report: Dict[str, Dict[str, Any]] = compute_robustness_metrics(
            baseline_dir=baseline_dir,
            trial_directories=trial_directories,
            asset_classes=canonical_order
        )

        # Step 3b: Record robustness report
        # We save the report in the base_output_path, alongside the trial directories
        report_path: Path = generate_robustness_report(
            metrics_report=metrics_report,
            output_dir=base_output_path
        )

    except Exception as exc:
        # Log the critical failure and raise a RuntimeError to halt the pipeline
        logger.critical(f"TASK 35 FAILED: {exc}. Halting robustness analysis.")
        raise RuntimeError(f"Robustness orchestration failed: {exc}") from exc

    logger.info(f"=== TASK 35 COMPLETED SUCCESSFULLY. Report at: {report_path} ===")

    # Return the path to the final report artifact
    return report_path
````

````python
# Task 36 — Robustness analysis: LLM variability and protocol stability (must use orchestrator)

# ==============================================================================
# Task 36: Robustness analysis: LLM variability and protocol stability
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 36, Step 1: Define the frozen perturbation schedule and generate configurations
# -------------------------------------------------------------------------------------------------------------------------------

def generate_llm_robustness_configurations(
    base_config: Dict[str, Any]
) -> List[Tuple[str, Dict[str, Any]]]:
    """
    Generates a suite of mutated configuration dictionaries for LLM variability
    analysis. Applies one-at-a-time protocol perturbations to reasoning effort
    schedules and web search capabilities.

    Parameters
    ----------
    base_config : Dict[str, Any]
        The validated, baseline study configuration dictionary.

    Returns
    -------
    List[Tuple[str, Dict[str, Any]]]
        A list of tuples, where each tuple contains a deterministic run ID string
        and the deeply copied, mutated configuration dictionary.

    Raises
    ------
    KeyError
        If the required baseline configuration keys are missing.
    """
    logger.info("Generating LLM robustness configurations for protocol stability analysis.")

    configurations: List[Tuple[str, Dict[str, Any]]] = []

    try:
        # Extract the frozen perturbation schedule from the configuration
        perturbation_configs: List[Dict[str, Any]] = base_config["REGISTRIES"]["ROBUSTNESS_LLM_VARIABILITY"]["perturbation_configs"]
    except KeyError as exc:
        raise KeyError(f"Base configuration is missing required keys for LLM perturbation: {exc}") from exc

    # Iterate through the prescribed perturbation schedule
    for p_config in perturbation_configs:
        run_id: str = p_config["run_id"]

        # Create a deep copy to ensure absolute state isolation between trials
        mutated_config: Dict[str, Any] = copy.deepcopy(base_config)

        # 1. Mutate Reasoning Effort Schedule
        effort_schedule: Dict[str, Any] = mutated_config["INFRASTRUCTURE"]["REASONING_EFFORT_SCHEDULE"]

        # CIO Agent Effort
        if "cio_agent" in effort_schedule:
            effort_schedule["cio_agent"]["effort"] = p_config["cio_reasoning_effort"]

        # CMA Judge Effort
        if "cma_judge" in effort_schedule:
            effort_schedule["cma_judge"]["effort"] = p_config["judge_reasoning_effort"]

        # Peer Review and Voting Effort
        if "peer_review_agent" in effort_schedule:
            effort_schedule["peer_review_agent"]["effort"] = p_config["peer_reasoning_effort"]
        if "voting_agent" in effort_schedule:
            effort_schedule["voting_agent"]["effort"] = p_config["peer_reasoning_effort"]

        # 2. Mutate Lookahead Controls (Web Search)
        lookahead_controls: Dict[str, Any] = mutated_config["INFRASTRUCTURE"]["LOOKAHEAD_CONTROLS"]
        lookahead_controls["web_search_default_enabled"] = p_config["web_search_enabled"]
        lookahead_controls["web_search_allowed_domains_baseline"] = p_config["web_search_domains"]

        # Append the mutated configuration to the output list
        configurations.append((run_id, mutated_config))
        logger.debug(f"Generated configuration for run ID: {run_id}")

    logger.info(f"Successfully generated {len(configurations)} LLM robustness configurations.")
    return configurations

# -------------------------------------------------------------------------------------------------------------------------------
# Task 36, Step 3: Measure and record stability of LLM outputs
# -------------------------------------------------------------------------------------------------------------------------------

def compute_llm_variability_metrics(
    baseline_dir: Path,
    trial_directories: Dict[str, Path],
    asset_classes: List[str],
    pc_methods: List[str]
) -> Dict[str, Dict[str, Any]]:
    """
    Computes strict mathematical and NLP-based drift metrics by comparing the
    artifacts of each perturbed LLM trial against the baseline run.

    CRITICAL REMEDIATION:
    1. Eradicates the TF-IDF sampling shortcut. Iterates over the entire
       assignment matrix (all possible reviewer-peer pairs) to compute the
       true mathematical mean of the cosine similarity across the full corpus.
    2. Eradicates blind array subtraction for the L1 distance. Enforces strict
       label alignment via dictionary mapping, identical to Task 35.

    Metrics computed:
    1. CMA Judge Selection Stability: Mean absolute difference in final estimates.
    2. Peer Review Sentiment Drift: Cosine similarity of TF-IDF vectorized text.
    3. Voting Outcome Stability: Spearman rank correlation of composite scores.
    4. Top-5 Overlap Fraction: Intersection of shortlists.
    5. Final Weight Drift: L1 distance of the final allocation vector.
    6. Ensemble Selection Changed: Boolean flag.

    Parameters
    ----------
    baseline_dir : Path
        The artifact directory of the unperturbed baseline run.
    trial_directories : Dict[str, Path]
        Dictionary mapping run IDs to their artifact directories.
    asset_classes : List[str]
        The canonical list of 18 asset classes.
    pc_methods : List[str]
        The canonical list of 21 PC methods.

    Returns
    -------
    Dict[str, Dict[str, Any]]
        A nested dictionary mapping run IDs to their computed drift metrics.
    """
    logger.info("Computing LLM variability metrics with full corpus traversal and strict label alignment.")

    metrics_report: Dict[str, Dict[str, Any]] = {}

    # --- Helper to safely load JSON ---
    def _load_json(path: Path) -> Dict[str, Any]:
        if not path.exists():
            return {}
        try:
            with open(path, "r", encoding="utf-8") as f:
                return json.load(f)
        except json.JSONDecodeError:
            return {}

    # --- Helper to safely load Text ---
    def _load_text(path: Path) -> str:
        if not path.exists():
            return ""
        try:
            with open(path, "r", encoding="utf-8") as f:
                return f.read()
        except OSError:
            return ""

    # --- Load Baseline Artifacts ---
    # 1. Baseline CMA Estimates
    base_cmas: Dict[str, float] = {}
    for ac in asset_classes:
        slug: str = __import__("re").sub(r"[^a-z0-9_]", "", ac.lower().replace(" ", "_"))
        cma_path_judge: Path = baseline_dir / "cma_judge" / slug / "cma.json"
        cma_path_ac: Path = baseline_dir / "ac_agents" / slug / "cma.json"
        cma_data: Dict[str, Any] = _load_json(cma_path_judge) if cma_path_judge.exists() else _load_json(cma_path_ac)
        base_cmas[ac] = float(cma_data.get("final_estimate", np.nan))

    # 2. Baseline Composite Scores
    base_ranking_data: Dict[str, Any] = _load_json(baseline_dir / "voting" / "composite_ranking.json")
    base_composite_scores: Dict[str, float] = base_ranking_data.get("composite_scores", {})

    # 3. Baseline Top-5
    base_top5_data: Dict[str, Any] = _load_json(baseline_dir / "top5_revision" / "top5_selection.json")
    base_top5: Set[str] = set(base_top5_data.get("top_5_methods", []))

    # 4. Baseline Final Weights and Labels (For strict alignment)
    base_weights_data: Dict[str, Any] = _load_json(baseline_dir / "cio" / "final_weights.json")
    base_weights_list: List[float] = base_weights_data.get("weights", [])
    base_labels_list: List[str] = base_weights_data.get("asset_class_labels", [])
    base_ensemble: str = base_weights_data.get("selected_ensemble", "UNKNOWN")

    base_weight_dict: Dict[str, float] = {}
    if len(base_weights_list) == 18 and len(base_labels_list) == 18:
        base_weight_dict = dict(zip(base_labels_list, base_weights_list))

    # --- Compute Metrics per Trial ---
    for run_id, trial_dir in trial_directories.items():
        trial_metrics: Dict[str, Any] = {}

        # 1. CMA Judge Selection Stability
        cma_drifts: List[float] = []
        for ac in asset_classes:
            slug: str = __import__("re").sub(r"[^a-z0-9_]", "", ac.lower().replace(" ", "_"))
            cma_path_judge: Path = trial_dir / "cma_judge" / slug / "cma.json"
            cma_path_ac: Path = trial_dir / "ac_agents" / slug / "cma.json"
            cma_data: Dict[str, Any] = _load_json(cma_path_judge) if cma_path_judge.exists() else _load_json(cma_path_ac)

            trial_est: float = float(cma_data.get("final_estimate", np.nan))
            base_est: float = base_cmas.get(ac, np.nan)

            if not np.isnan(trial_est) and not np.isnan(base_est):
                cma_drifts.append(abs(trial_est - base_est))

        trial_metrics["cma_judge_selection_changes"] = float(np.mean(cma_drifts)) if cma_drifts else np.nan

        # 2. Peer Review Sentiment Drift (NLP)
        # CRITICAL REMEDIATION: Full corpus traversal, no sampling shortcuts.
        if _SKLEARN_AVAILABLE:
            similarities: List[float] = []
            vectorizer = TfidfVectorizer(stop_words='english')

            # Iterate through all possible reviewer-peer combinations
            for reviewer in pc_methods:
                for peer in pc_methods:
                    if reviewer == peer:
                        continue # Self-reviews do not exist

                    filename: str = f"{reviewer}_reviews_{peer}.md"
                    base_text: str = _load_text(baseline_dir / "peer_reviews" / filename)
                    trial_text: str = _load_text(trial_dir / "peer_reviews" / filename)

                    # Only compute similarity if the review exists in both the baseline and the trial
                    if base_text and trial_text:
                        try:
                            tfidf_matrix = vectorizer.fit_transform([base_text, trial_text])
                            sim: float = float(cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0])
                            similarities.append(sim)
                        except ValueError:
                            # Catch ValueError if the vocabulary is empty (e.g., only stop words present)
                            pass

            trial_metrics["peer_review_cosine_similarity"] = float(np.mean(similarities)) if similarities else np.nan
        else:
            trial_metrics["peer_review_cosine_similarity"] = np.nan
            logger.warning("scikit-learn not available. Skipping NLP drift metrics.")

        # 3. Voting Outcome Stability (Spearman Rank Correlation)
        trial_ranking_data: Dict[str, Any] = _load_json(trial_dir / "voting" / "composite_ranking.json")
        trial_composite_scores: Dict[str, float] = trial_ranking_data.get("composite_scores", {})

        base_scores_list: List[float] = []
        trial_scores_list: List[float] = []

        for method in pc_methods:
            if method in base_composite_scores and method in trial_composite_scores:
                base_scores_list.append(base_composite_scores[method])
                trial_scores_list.append(trial_composite_scores[method])

        if len(base_scores_list) > 1:
            rho, _ = spearmanr(base_scores_list, trial_scores_list)
            trial_metrics["composite_ranking_spearman_rho"] = float(rho)
        else:
            trial_metrics["composite_ranking_spearman_rho"] = np.nan

        # 4. Top-5 Overlap Fraction
        trial_top5_data: Dict[str, Any] = _load_json(trial_dir / "top5_revision" / "top5_selection.json")
        trial_top5: Set[str] = set(trial_top5_data.get("top_5_methods", []))
        if base_top5 and trial_top5:
            overlap_fraction: float = len(base_top5.intersection(trial_top5)) / 5.0
        else:
            overlap_fraction = np.nan
        trial_metrics["top5_overlap_fraction"] = float(overlap_fraction)

        # 5. Final Weight Drift (L1 Distance with Strict Alignment)
        # CRITICAL REMEDIATION: Enforce label alignment before vector arithmetic
        trial_weights_data: Dict[str, Any] = _load_json(trial_dir / "cio" / "final_weights.json")
        trial_weights_list: List[float] = trial_weights_data.get("weights", [])
        trial_labels_list: List[str] = trial_weights_data.get("asset_class_labels", [])

        if base_weight_dict and len(trial_weights_list) == 18 and len(trial_labels_list) == 18:
            if set(base_labels_list) == set(trial_labels_list):
                trial_weight_dict: Dict[str, float] = dict(zip(trial_labels_list, trial_weights_list))

                aligned_base_arr: np.ndarray = np.array([base_weight_dict[ac] for ac in asset_classes], dtype=np.float64)
                aligned_trial_arr: np.ndarray = np.array([trial_weight_dict[ac] for ac in asset_classes], dtype=np.float64)

                l1_dist: float = float(np.sum(np.abs(aligned_trial_arr - aligned_base_arr)))
                trial_metrics["final_weight_l1_distance"] = round(l1_dist, 6)
            else:
                logger.error(f"Label mismatch in trial {run_id}. Cannot compute L1 distance safely.")
                trial_metrics["final_weight_l1_distance"] = np.nan
        else:
            trial_metrics["final_weight_l1_distance"] = np.nan

        # 6. Ensemble Selection Changed
        trial_ensemble: str = trial_weights_data.get("selected_ensemble", "UNKNOWN_TRIAL")
        trial_metrics["ensemble_selection_changed"] = bool(trial_ensemble != base_ensemble)
        trial_metrics["trial_ensemble"] = trial_ensemble

        # 7. IPS Compliance
        trial_metrics["ips_compliant"] = bool(trial_weights_data.get("range_constraint_verified", False))

        metrics_report[run_id] = trial_metrics
        logger.debug(f"Computed LLM variability metrics for {run_id}: Rho={trial_metrics['composite_ranking_spearman_rho']}, L1={trial_metrics['final_weight_l1_distance']}")

    return metrics_report

# -------------------------------------------------------------------------------------------------------------------------------
# Task 36, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_36_llm_variability_robustness(
    base_config: Dict[str, Any],
    raw_dataframes: Dict[str, pd.DataFrame],
    client: Any,
    baseline_dir: Path,
    base_output_path: Path
) -> Path:
    """
    The Master Orchestrator for Task 36. Executes the full LLM variability and
    protocol stability analysis pipeline.

    Generates perturbed configurations, executes isolated end-to-end trials,
    computes mathematical and NLP drift metrics against the baseline, and
    synthesizes the findings into a structured fiduciary report.

    Parameters
    ----------
    base_config : Dict[str, Any]
        The validated, baseline study configuration dictionary.
    raw_dataframes : Dict[str, pd.DataFrame]
        The raw, unmutated dataframes required for pipeline ingestion.
    client : Any
        The authenticated OpenAI client.
    baseline_dir : Path
        The artifact directory of the unperturbed baseline run. Must exist.
    base_output_path : Path
        The root directory where the robustness trial folders will be created.

    Returns
    -------
    Path
        The absolute path to the final robustness_llm_variability_report.json artifact.

    Raises
    ------
    FileNotFoundError
        If the baseline_dir does not exist, making comparison impossible.
    RuntimeError
        If any unhandled exception occurs during the orchestration flow,
        enforcing a strict fail-closed halt.
    """
    logger.info("=== STARTING TASK 36: LLM VARIABILITY AND PROTOCOL STABILITY ANALYSIS ===")

    # Pre-flight check: Ensure the baseline directory exists for comparison
    if not baseline_dir.exists() or not baseline_dir.is_dir():
        raise FileNotFoundError(
            f"Fatal Error: Baseline directory {baseline_dir} does not exist. "
            f"Robustness analysis requires a completed baseline run for comparison."
        )

    # Extract the canonical asset classes and PC methods for drift computation
    try:
        asset_classes_dict: Dict[str, List[str]] = base_config["IPS_GOVERNANCE"]["ASSET_CLASSES"]
        canonical_order: List[str] = []
        for category in ["equity", "fixed_income", "real_assets", "cash"]:
            canonical_order.extend(asset_classes_dict.get(category, []))

        pc_registry: Dict[str, Any] = base_config["REGISTRIES"]["PC_AGENTS"]["method_configs"]
        pc_methods: List[str] = list(pc_registry.keys())
    except KeyError as exc:
        raise RuntimeError(f"Configuration is missing required keys for Task 36: {exc}") from exc

    # Wrap the DAG in a try/except block to guarantee fail-closed semantics
    try:
        # Step 1: Generate alternative frozen parameter sets
        configurations: List[Tuple[str, Dict[str, Any]]] = generate_llm_robustness_configurations(
            base_config=base_config
        )

        # Step 2: Execute isolated robustness trials
        # We reuse the robust executor from Task 35
        trial_directories: Dict[str, Path] = execute_robustness_trials(
            configurations=configurations,
            raw_dataframes=raw_dataframes,
            client=client,
            base_output_path=base_output_path
        )

        # Guard: Ensure at least one trial succeeded before attempting metric computation
        if not trial_directories:
            raise RuntimeError("All LLM robustness trials failed. Cannot compute drift metrics.")

        # Step 3: Compute mathematical and NLP drift metrics
        metrics_report: Dict[str, Dict[str, Any]] = compute_llm_variability_metrics(
            baseline_dir=baseline_dir,
            trial_directories=trial_directories,
            asset_classes=canonical_order,
            pc_methods=pc_methods
        )

        # Step 4: Record robustness report
        # We reuse the report generator from Task 35, saving it with a specific name
        report_path: Path = generate_robustness_report(
            metrics_report=metrics_report,
            output_dir=base_output_path
        )

        # Rename the report to match the Task 36 specification
        final_report_path: Path = base_output_path / "robustness_llm_variability_report.json"
        report_path.rename(final_report_path)

    except Exception as exc:
        # Log the critical failure and raise a RuntimeError to halt the pipeline
        logger.critical(f"TASK 36 FAILED: {exc}. Halting LLM variability analysis.")
        raise RuntimeError(f"LLM Robustness orchestration failed: {exc}") from exc

    logger.info(f"=== TASK 36 COMPLETED SUCCESSFULLY. Report at: {final_report_path} ===")

    # Return the path to the final report artifact
    return final_report_path
````

````python
# Task 37 — Final reproducibility package (audit-ready deliverable)

# ==============================================================================
# Task 37: Final reproducibility package (audit-ready deliverable)
# ==============================================================================

# -------------------------------------------------------------------------------------------------------------------------------
# Task 37, Step 2: Persist provenance (Cryptographic linkage)
# -------------------------------------------------------------------------------------------------------------------------------

# Note: Executed before Step 1 so the provenance file is included in the archive.
def persist_provenance(
    raw_dataframes: Dict[str, pd.DataFrame],
    config: Dict[str, Any],
    run_id: str,
    run_output_dir: Path
) -> Dict[str, Any]:
    """
    Cryptographically links the inputs to the outputs by computing SHA-256 hashes
    of all raw dataframes and the configuration dictionary. Persists this metadata
    to provenance.json.

    Parameters
    ----------
    raw_dataframes : Dict[str, pd.DataFrame]
        The dictionary of raw, unmutated dataframes ingested by the pipeline.
    config : Dict[str, Any]
        The validated study configuration dictionary.
    run_id : str
        The unique identifier for the current run.
    run_output_dir : Path
        The root directory for the current run's artifacts.

    Returns
    -------
    Dict[str, Any]
        The constructed provenance dictionary.

    Raises
    ------
    OSError
        If the provenance.json file cannot be written to disk.
    """
    logger.info("Computing cryptographic provenance hashes for all inputs.")

    provenance_data: Dict[str, Any] = {
        "run_id": run_id,
        "timestamp_utc": datetime.now(tz=timezone.utc).isoformat(),
        "data_hashes_sha256": {},
        "config_hash_sha256": "",
        "seeds": config.get("RANDOM_SEEDS", {}),
        "llm_policy": config.get("INFRASTRUCTURE", {}).get("LLM_INVOCATION_POLICY", {})
    }

    # 1. Hash Raw Dataframes
    # We hash the in-memory state to guarantee the hash reflects the exact data used
    for df_name, df in raw_dataframes.items():
        try:
            # Serialize to CSV byte stream for consistent hashing
            csv_bytes: bytes = df.to_csv(index=True).encode('utf-8')
            df_hash: str = hashlib.sha256(csv_bytes).hexdigest()
            provenance_data["data_hashes_sha256"][df_name] = df_hash
        except Exception as exc:
            logger.warning(f"Failed to compute hash for dataframe '{df_name}': {exc}")
            provenance_data["data_hashes_sha256"][df_name] = "HASH_COMPUTATION_FAILED"

    # 2. Hash Configuration
    try:
        # Serialize config to a canonical JSON string (sorted keys) for deterministic hashing
        config_json_bytes: bytes = json.dumps(config, sort_keys=True, default=str).encode('utf-8')
        config_hash: str = hashlib.sha256(config_json_bytes).hexdigest()
        provenance_data["config_hash_sha256"] = config_hash
    except Exception as exc:
        logger.warning(f"Failed to compute hash for configuration: {exc}")
        provenance_data["config_hash_sha256"] = "HASH_COMPUTATION_FAILED"

    # 3. Persist to Disk
    provenance_path: Path = run_output_dir / "provenance.json"
    try:
        with open(provenance_path, "w", encoding="utf-8") as f:
            json.dump(provenance_data, f, indent=4)
    except OSError as exc:
        raise OSError(f"Failed to write provenance.json to {provenance_path}: {exc}") from exc

    logger.info(f"Successfully persisted cryptographic provenance to {provenance_path}.")
    return provenance_data

# -------------------------------------------------------------------------------------------------------------------------------
# Task 37, Step 3: Produce a single "Reproduction Report"
# Note: Executed before Step 1 so the report is included in the archive.
# -------------------------------------------------------------------------------------------------------------------------------

def produce_reproduction_report(
    run_id: str,
    config: Dict[str, Any],
    run_output_dir: Path
) -> Path:
    """
    Programmatically generates a markdown report explicitly documenting all
    methodological gap-filling and "UNSPECIFIED IN MANUSCRIPT" choices that
    were frozen in the configuration to enable deterministic execution.

    Parameters
    ----------
    run_id : str
        The unique identifier for the current run.
    config : Dict[str, Any]
        The validated study configuration dictionary.
    run_output_dir : Path
        The root directory for the current run's artifacts.

    Returns
    -------
    Path
        The absolute path to the written reproduction_report.md file.

    Raises
    ------
    OSError
        If the report cannot be written to disk.
    """
    logger.info("Generating methodological reproduction report.")

    # Extract relevant configuration sections
    data_conv: Dict[str, Any] = config.get("DATA_CONVENTIONS", {})
    infra: Dict[str, Any] = config.get("INFRASTRUCTURE", {})
    method_params: Dict[str, Any] = config.get("METHODOLOGY_PARAMS", {})

    md_lines: List[str] = [
        "# Self-Driving Portfolio: Reproduction & Methodology Report",
        f"**Run ID:** `{run_id}`",
        f"**Timestamp (UTC):** `{datetime.now(tz=timezone.utc).isoformat()}`",
        "",
        "## 1. Purpose",
        "This document explicitly records the methodological assumptions and gap-filling required to translate the high-level architecture described in Ang, Azimbayev, and Kim (2026) into a deterministic, executable pipeline. Items marked as 'UNSPECIFIED IN MANUSCRIPT' have been frozen to the values below to guarantee reproducibility.",
        "",
        "## 2. Data Vintage & Cleansing Policies",
        f"- **Point-in-Time Enforcement:** {data_conv.get('point_in_time_policy', {}).get('valuation_as_of_enforced', True)}",
        f"- **GDP Vintage Policy:** {data_conv.get('point_in_time_policy', {}).get('gdp_vintage_policy', 'latest_available_with_approximation_disclaimer')}",
        f"- **Missing Data Forward-Fill Limit:** {data_conv.get('missing_data_policy', {}).get('forward_fill_limit', 5)} periods",
        "",
        "## 3. Engineered Methodological Assumptions (Gap-Filling)",
        f"- **Return Formula:** `{data_conv.get('return_formula', 'simple_periodic')}` (Required for cross-sectional aggregation; log-returns rejected).",
        f"- **Annualization Multipliers:** Return = `{data_conv.get('annualisation', {}).get('mu_multiplier', 12)}`, Volatility = `{data_conv.get('annualisation', {}).get('sigma_multiplier', 'sqrt(12)')}`.",
        f"- **Covariance Estimator:** `{method_params.get('COVARIANCE_ESTIMATION', {}).get('estimator', 'SHRINKAGE_LEDOIT_WOLF')}` (Required to prevent singular matrices).",
        f"- **Covariance Window:** `{method_params.get('COVARIANCE_ESTIMATION', {}).get('window_months', 120)}` months.",
        f"- **Maximum Drawdown Window:** Expanding window from inception to $t$.",
        f"- **Rebalancing Assumption for Backtest:** Static weights between rebalances.",
        f"- **Risk Contribution Formula:** Variance-based marginal contribution (Euler).",
        "",
        "## 4. Agent Framework Assignments",
        "The manuscript describes agentic interactions but does not prescribe specific orchestration frameworks per task. The following assignments were engineered for this implementation:",
    ]

    # Append framework assignments
    frameworks: Dict[str, Dict[str, str]] = infra.get("AGENT_FRAMEWORK_ASSIGNMENTS", {})
    for task, details in frameworks.items():
        md_lines.append(f"- **{task}:** `{details.get('framework', 'Unknown')}` - *{details.get('rationale', '')}*")

    md_content: str = "\n".join(md_lines)

    report_path: Path = run_output_dir / "reproduction_report.md"

    try:
        with open(report_path, "w", encoding="utf-8") as f:
            f.write(md_content)
    except OSError as exc:
        raise OSError(f"Failed to write reproduction report to {report_path}: {exc}") from exc

    logger.info(f"Successfully generated reproduction report at {report_path}.")
    return report_path.resolve()

# -------------------------------------------------------------------------------------------------------------------------------
# Task 37, Step 1: Persist all artifacts (Directory Compression)
# -------------------------------------------------------------------------------------------------------------------------------

def persist_all_artifacts(run_output_dir: Path) -> Path:
    """
    Compiles the entire output directory tree into a distributable, compressed
    archive format (.zip), ensuring the fiduciary record is immutable and easily
    transferable.

    Parameters
    ----------
    run_output_dir : Path
        The root directory containing all artifacts generated during the run.

    Returns
    -------
    Path
        The absolute path to the compressed archive file.

    Raises
    ------
    OSError
        If the directory cannot be archived due to filesystem permissions or locks.
    ValueError
        If the resulting archive file is empty (zero bytes).
    """
    logger.info(f"Compressing artifact directory: {run_output_dir}")

    # Define the base name for the archive, matching the name of the output directory
    archive_base_name: str = str(run_output_dir.resolve())

    # Attempt to compress the entire output directory into a ZIP archive
    try:
        # shutil.make_archive safely packages the directory structure
        archive_path_str: str = shutil.make_archive(
            base_name=archive_base_name,
            format="zip",
            root_dir=str(run_output_dir.parent),
            base_dir=run_output_dir.name
        )
    except Exception as exc:
        # Raise a fatal OSError if the immutable deliverable cannot be created
        raise OSError(f"Failed to create compressed archive for {run_output_dir}: {exc}") from exc

    archive_path: Path = Path(archive_path_str)

    # Completeness Verification: Assert that the archive exists and is not empty
    if not archive_path.exists():
        raise OSError(f"Archive creation reported success, but file not found at {archive_path}.")

    if archive_path.stat().st_size == 0:
        raise ValueError(f"Fatal Error: Created archive {archive_path} is empty (0 bytes).")

    logger.info(f"Successfully created immutable artifact archive at: {archive_path} ({archive_path.stat().st_size} bytes)")
    return archive_path.resolve()

# -------------------------------------------------------------------------------------------------------------------------------
# Task 37, Orchestrator Function
# -------------------------------------------------------------------------------------------------------------------------------

def orchestrate_task_37_final_reproducibility_package(
    raw_dataframes: Dict[str, pd.DataFrame],
    config: Dict[str, Any],
    run_id: str,
    run_output_dir: Path
) -> Path:
    """
    Orchestrates the final reproducibility package generation. Computes
    cryptographic provenance, generates the methodological report, and
    compresses the entire artifact directory into an immutable deliverable.

    Parameters
    ----------
    raw_dataframes : Dict[str, pd.DataFrame]
        The dictionary of raw, unmutated dataframes ingested by the pipeline.
    config : Dict[str, Any]
        The validated study configuration dictionary.
    run_id : str
        The unique identifier for the current run.
    run_output_dir : Path
        The root directory for the current run's artifacts.

    Returns
    -------
    Path
        The absolute path to the final compressed archive file (.zip).

    Raises
    ------
    RuntimeError
        If any step of the finalization process fails, enforcing fail-closed semantics.
    """
    logger.info("=== STARTING TASK 37: FINAL REPRODUCIBILITY PACKAGE ===")

    try:
        # Step 2: Persist provenance (Cryptographic linkage)
        # Executed first so the provenance.json is included in the final zip
        persist_provenance(
            raw_dataframes=raw_dataframes,
            config=config,
            run_id=run_id,
            run_output_dir=run_output_dir
        )

        # Step 3: Produce a single "Reproduction Report"
        # Executed second so the reproduction_report.md is included in the final zip
        produce_reproduction_report(
            run_id=run_id,
            config=config,
            run_output_dir=run_output_dir
        )

        # Step 1: Persist all artifacts (Directory Compression)
        # Executed last to seal the directory
        archive_path: Path = persist_all_artifacts(
            run_output_dir=run_output_dir
        )

    except Exception as exc:
        logger.critical(f"TASK 37 FAILED: {exc}. The reproducibility package could not be finalized.")
        raise RuntimeError(f"Final reproducibility package generation failed: {exc}") from exc

    logger.info(f"=== TASK 37 COMPLETED SUCCESSFULLY. Deliverable at: {archive_path} ===")
    return archive_path
````

````python
# Top Level Orchestrator

# ==============================================================================
# Master Orchestrator: The Fiduciary Intelligence Cycle
# ==============================================================================

def orchestrate_fiduciary_intelligence_cycle(
    raw_config: Dict[str, Any],
    raw_dataframes: Dict[str, pd.DataFrame],
    client: Any,
    base_output_path: str = "./run_outputs"
) -> Dict[str, Optional[Path]]:
    """
    The apex orchestrator for the "Self-Driving Portfolio" architecture.

    Executes the complete fiduciary intelligence cycle in a strict, sequential
    Directed Acyclic Graph (DAG). This includes the baseline end-to-end run,
    the data and modeling sensitivity analysis, the LLM variability analysis,
    and the final cryptographic packaging of the entire execution state.

    CRITICAL REMEDIATION: Implements Master Cycle Encapsulation to prevent
    directory pollution. Replaces ungraceful sys.exit(1) calls with Pythonic
    RuntimeError exceptions to enforce fail-closed semantics while adhering
    to professional software engineering standards.

    Parameters
    ----------
    raw_config : Dict[str, Any]
        The raw, unvalidated STUDY_CONFIG dictionary.
    raw_dataframes : Dict[str, pd.DataFrame]
        A dictionary mapping dataframe string identifiers to their raw pandas DataFrames.
    client : Any
        The authenticated OpenAI client instance.
    base_output_path : str, optional
        The root path where the master cycle directory will be created.
        Defaults to "./run_outputs".

    Returns
    -------
    Dict[str, Optional[Path]]
        A dictionary containing the absolute paths to the generated artifacts:
        - 'baseline_archive_path': The zipped baseline run.
        - 'data_robustness_report_path': The JSON report from Task 35 (if successful).
        - 'llm_variability_report_path': The JSON report from Task 36 (if successful).
        - 'master_deliverable_path': The zipped archive of the entire master cycle directory.

    Raises
    ------
    RuntimeError
        If the baseline run (Task 34) fails, or if the final packaging (Task 37) fails,
        enforcing a strict fail-closed halt.
    """
    logger.info("=====================================================================")
    logger.info("INITIATING THE FIDUCIARY INTELLIGENCE CYCLE")
    logger.info("=====================================================================")

    # -------------------------------------------------------------------------------------------
    # REMEDIATION STEP 1: Master Cycle Encapsulation
    # -------------------------------------------------------------------------------------------
    # Generate a cryptographically random UUID to ensure the cycle directory is globally unique
    run_uuid: str = str(uuid.uuid4())

    # Format the timestamp to be filesystem-safe (removing colons and special characters)
    safe_timestamp: str = datetime.now(tz=timezone.utc).strftime("%Y%m%d_%H%M%S")

    # Construct the unique directory name combining the timestamp and the UUID
    cycle_dir_name: str = f"cycle_{safe_timestamp}_{run_uuid[:8]}"

    # Resolve the absolute path for the unique master cycle directory
    master_cycle_dir: Path = Path(base_output_path).resolve() / cycle_dir_name

    # Create the master cycle directory, including any necessary parent directories
    master_cycle_dir.mkdir(parents=True, exist_ok=True)

    logger.info(f"Master cycle directory established at: {master_cycle_dir}")

    # Initialize the output dictionary to track generated artifacts
    cycle_artifacts: Dict[str, Optional[Path]] = {
        "baseline_archive_path": None,
        "data_robustness_report_path": None,
        "llm_variability_report_path": None,
        "master_deliverable_path": None
    }

    # -------------------------------------------------------------------------------------------
    # STEP 1: Execute the Baseline End-to-End Pipeline (Task 34)
    # -------------------------------------------------------------------------------------------
    logger.info("--- STEP 1: Executing Baseline End-to-End Pipeline ---")
    try:
        # The baseline run is the anchor for all subsequent robustness trials.
        # REMEDIATION STEP 2: Strict Path Routing. We pass master_cycle_dir as the base_output_path.
        baseline_archive_path: Path = orchestrate_task_34_end_to_end_pipeline(
            raw_config=raw_config,
            raw_dataframes=raw_dataframes,
            client=client,
            base_output_path=str(master_cycle_dir)
        )
        cycle_artifacts["baseline_archive_path"] = baseline_archive_path
        logger.info(f"Baseline pipeline completed successfully. Archive at: {baseline_archive_path}")

    except Exception as exc:
        # REMEDIATION STEP 4: Pythonic Fail-Closed Semantics
        logger.critical(f"FATAL ERROR: Baseline pipeline execution failed: {exc}")
        logger.critical("The fiduciary intelligence cycle cannot proceed without a baseline. Halting.")
        raise RuntimeError(f"Baseline pipeline execution failed: {exc}") from exc

    # -------------------------------------------------------------------------------------------
    # STEP 2: Extract Baseline Directory Path
    # -------------------------------------------------------------------------------------------
    # Task 34 returns the path to the .zip archive. We must derive the uncompressed
    # directory path to serve as the comparison anchor for Tasks 35 and 36.
    # By stripping the '.zip' suffix, we obtain the original directory path.
    baseline_dir: Path = baseline_archive_path.with_suffix('')

    if not baseline_dir.exists() or not baseline_dir.is_dir():
        # REMEDIATION STEP 4: Pythonic Fail-Closed Semantics
        logger.critical(f"FATAL ERROR: Derived baseline directory does not exist at {baseline_dir}.")
        raise RuntimeError(f"Derived baseline directory does not exist at {baseline_dir}.")

    # -------------------------------------------------------------------------------------------
    # STEP 3: Execute Data and Modeling Robustness Analysis (Task 35)
    # -------------------------------------------------------------------------------------------
    logger.info("--- STEP 3: Executing Data and Modeling Robustness Analysis ---")
    try:
        # We pass a deep copy of the config to ensure the baseline state is not mutated
        # REMEDIATION STEP 2: Strict Path Routing. We pass master_cycle_dir as the base_output_path.
        data_robustness_report_path: Path = orchestrate_task_35_data_modeling_robustness(
            base_config=copy.deepcopy(raw_config),
            raw_dataframes=raw_dataframes,
            client=client,
            baseline_dir=baseline_dir,
            base_output_path=master_cycle_dir
        )
        cycle_artifacts["data_robustness_report_path"] = data_robustness_report_path
        logger.info(f"Data robustness analysis completed. Report at: {data_robustness_report_path}")

    except Exception as exc:
        # Graceful degradation: If sensitivity analysis fails, log the error but continue
        # to ensure the baseline run is still packaged and delivered.
        logger.error(f"Data robustness analysis failed: {exc}. Proceeding to next step.")

    # -------------------------------------------------------------------------------------------
    # STEP 4: Execute LLM Variability and Protocol Stability Analysis (Task 36)
    # -------------------------------------------------------------------------------------------
    logger.info("--- STEP 4: Executing LLM Variability and Protocol Stability Analysis ---")
    try:
        # We pass a deep copy of the config to ensure the baseline state is not mutated
        # REMEDIATION STEP 2: Strict Path Routing. We pass master_cycle_dir as the base_output_path.
        llm_variability_report_path: Path = orchestrate_task_36_llm_variability_robustness(
            base_config=copy.deepcopy(raw_config),
            raw_dataframes=raw_dataframes,
            client=client,
            baseline_dir=baseline_dir,
            base_output_path=master_cycle_dir
        )
        cycle_artifacts["llm_variability_report_path"] = llm_variability_report_path
        logger.info(f"LLM variability analysis completed. Report at: {llm_variability_report_path}")

    except Exception as exc:
        # Graceful degradation: Log the error but continue to final packaging.
        logger.error(f"LLM variability analysis failed: {exc}. Proceeding to final packaging.")

    # -------------------------------------------------------------------------------------------
    # STEP 5: Final Reproducibility Package (Task 37)
    # -------------------------------------------------------------------------------------------
    logger.info("--- STEP 5: Generating Master Reproducibility Package ---")
    try:
        # REMEDIATION STEP 3: Targeted Cryptographic Sealing
        # Task 37 cryptographically seals the entire master_cycle_dir, which now contains
        # the baseline run, all trial runs, and the robustness reports.
        # We use the master cycle directory's name as the master run ID.
        master_run_id: str = master_cycle_dir.name

        master_deliverable_path: Path = orchestrate_task_37_final_reproducibility_package(
            raw_dataframes=raw_dataframes,
            config=raw_config,
            run_id=master_run_id,
            run_output_dir=master_cycle_dir
        )
        cycle_artifacts["master_deliverable_path"] = master_deliverable_path
        logger.info(f"Master reproducibility package created at: {master_deliverable_path}")

    except Exception as exc:
        # REMEDIATION STEP 4: Pythonic Fail-Closed Semantics
        logger.critical(f"FATAL ERROR: Failed to generate master reproducibility package: {exc}")
        raise RuntimeError(f"Failed to generate master reproducibility package: {exc}") from exc

    logger.info("=====================================================================")
    logger.info("FIDUCIARY INTELLIGENCE CYCLE COMPLETED SUCCESSFULLY")
    logger.info("=====================================================================")

    return cycle_artifacts
````
